```python
import os
import json
import threading
import tkinter as tk
from tkinter import ttk, messagebox, scrolledtext
from glob import glob
from datetime import datetime
from pylsl import resolve_streams
from liesl import LabRecorder

class LSLRecorderApp:
    def __init__(self, root):
        self.root = root
        self.root.title("LSL Stream Recorder")
        self.root.geometry("500x450")
        self.root.protocol("WM_DELETE_WINDOW", self.on_closing)
        
        # Configuration
        self.METADATA_FOLDER = r"C:\Users\cogpsy-vrlab\Documents\GitHub\BreathingSpace\Level2_RunExperiment\Results\participant_metadata"
        self.OUTPUT_FOLDER = r"C:\Users\cogpsy-vrlab\Documents\GitHub\BreathingSpace\Level2_RunExperiment\Results\LSL_Output"
        
        # Ensure output folder exists
        os.makedirs(self.OUTPUT_FOLDER, exist_ok=True)
        
        # Variables
        self.recorder = None
        self.recording = False
        self.participant_id = tk.StringVar(value="None")
        self.status_var = tk.StringVar(value="Ready")
        self.recording_file_var = tk.StringVar(value="Not recording")
        self.streams_var = tk.StringVar(value="No streams found")
        
        # Create UI
        self.create_ui()
        
        # Try to get latest participant ID on startup
        self.refresh_participant_id()
    
    def create_ui(self):
        # Main frame
        main_frame = ttk.Frame(self.root, padding=10)
        main_frame.pack(fill=tk.BOTH, expand=True)
        
        # Title
        ttk.Label(main_frame, text="LSL Stream Recorder", font=("Arial", 16, "bold")).pack(pady=10)
        
        # Participant frame
        participant_frame = ttk.LabelFrame(main_frame, text="Participant", padding=10)
        participant_frame.pack(fill=tk.X, padx=5, pady=5)
        
        # Participant ID display
        id_frame = ttk.Frame(participant_frame)
        id_frame.pack(fill=tk.X, pady=5)
        
        ttk.Label(id_frame, text="Participant ID:").pack(side=tk.LEFT, padx=5)
        ttk.Label(id_frame, textvariable=self.participant_id, font=("Arial", 10, "bold")).pack(side=tk.LEFT, padx=5)
        ttk.Button(id_frame, text="Refresh", command=self.refresh_participant_id).pack(side=tk.RIGHT, padx=5)
        
        # Recording controls
        controls_frame = ttk.LabelFrame(main_frame, text="Recording Controls", padding=10)
        controls_frame.pack(fill=tk.X, padx=5, pady=5)
        
        # Button frame
        button_frame = ttk.Frame(controls_frame)
        button_frame.pack(fill=tk.X, pady=5)
        
        self.start_button = ttk.Button(button_frame, text="START RECORDING", command=self.start_recording, width=20)
        self.start_button.pack(side=tk.LEFT, padx=5)
        
        self.stop_button = ttk.Button(button_frame, text="STOP RECORDING", command=self.stop_recording, width=20, state=tk.DISABLED)
        self.stop_button.pack(side=tk.LEFT, padx=5)
        
        # Status frame
        status_frame = ttk.LabelFrame(main_frame, text="Status", padding=10)
        status_frame.pack(fill=tk.X, padx=5, pady=5)
        
        ttk.Label(status_frame, text="Status:").grid(row=0, column=0, sticky=tk.W, padx=5, pady=2)
        ttk.Label(status_frame, textvariable=self.status_var, font=("Arial", 10, "bold")).grid(row=0, column=1, sticky=tk.W, padx=5, pady=2)
        
        ttk.Label(status_frame, text="Recording File:").grid(row=1, column=0, sticky=tk.W, padx=5, pady=2)
        ttk.Label(status_frame, textvariable=self.recording_file_var).grid(row=1, column=1, sticky=tk.W, padx=5, pady=2)
        
        ttk.Label(status_frame, text="Streams:").grid(row=2, column=0, sticky=tk.W, padx=5, pady=2)
        ttk.Label(status_frame, textvariable=self.streams_var).grid(row=2, column=1, sticky=tk.W, padx=5, pady=2)
        
        # Log area
        log_frame = ttk.LabelFrame(main_frame, text="Log", padding=10)
        log_frame.pack(fill=tk.BOTH, expand=True, padx=5, pady=5)
        
        self.log_text = scrolledtext.ScrolledText(log_frame, height=8, wrap=tk.WORD)
        self.log_text.pack(fill=tk.BOTH, expand=True)
    
    def log(self, message):
        """Add message to log with timestamp"""
        timestamp = datetime.now().strftime("%H:%M:%S")
        log_message = f"[{timestamp}] {message}\n"
        self.log_text.insert(tk.END, log_message)
        self.log_text.see(tk.END)
        print(log_message.strip())
    
    def refresh_participant_id(self):
        """Get the most recent participant ID from JSON files"""
        try:
            json_files = glob(os.path.join(self.METADATA_FOLDER, "*.json"))
            
            if not json_files:
                self.log("No participant JSON files found.")
                self.participant_id.set("None")
                return
            
            # Get most recently modified file
            latest_file = max(json_files, key=os.path.getmtime)
            
            with open(latest_file, "r") as f:
                metadata = json.load(f)
            
            if "participant_id" not in metadata:
                self.log(f"participant_id missing in {os.path.basename(latest_file)}")
                self.participant_id.set("Error")
                return
            
            participant_id = metadata["participant_id"]
            self.participant_id.set(participant_id)
            self.log(f"Found participant: {participant_id}")
            
        except Exception as e:
            self.log(f"Error refreshing participant ID: {str(e)}")
            self.participant_id.set("Error")
    
    def start_recording(self):
        """Start recording LSL streams"""
        if self.recording:
            messagebox.showinfo("Already Recording", "Recording is already in progress.")
            return
        
        # Check if participant ID is available
        if self.participant_id.get() in ["None", "Error"]:
            messagebox.showerror("Error", "No valid participant ID found. Please refresh or check metadata folder.")
            return
        
        # Start recording in a separate thread
        threading.Thread(target=self._recording_thread, daemon=True).start()
    
    def _recording_thread(self):
        """Background thread for recording to prevent UI freezing"""
        try:
            # Update UI state
            self.root.after(0, lambda: self.status_var.set("Resolving streams..."))
            self.root.after(0, lambda: self.start_button.config(state=tk.DISABLED))
            
            # Find available streams
            streams = resolve_streams()
            
            if not streams:
                self.root.after(0, lambda: self.log("No LSL streams found. Cannot start recording."))
                self.root.after(0, lambda: self.status_var.set("No streams available"))
                self.root.after(0, lambda: self.start_button.config(state=tk.NORMAL))
                return
            
            # Update streams info
            stream_info = f"{len(streams)} streams found"
            self.root.after(0, lambda: self.streams_var.set(stream_info))
            self.root.after(0, lambda: self.log(f"Found {len(streams)} LSL streams"))
            
            # Prepare the output file path
            participant_id = self.participant_id.get()
            timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
            output_filename = os.path.join(self.OUTPUT_FOLDER, f"{participant_id}_{timestamp}.xdf")
            
            # Create recorder and start recording
            self.recorder = LabRecorder()
            self.recorder.start_recording(
                filename=output_filename,
                streamargs=[{"type": stream.type()} for stream in streams]
            )
            
            # Update status
            self.recording = True
            self.root.after(0, lambda: self.status_var.set("Recording in progress"))
            self.root.after(0, lambda: self.recording_file_var.set(os.path.basename(output_filename)))
            self.root.after(0, lambda: self.stop_button.config(state=tk.NORMAL))
            self.root.after(0, lambda: self.log(f"Recording started to {output_filename}"))
            
        except Exception as e:
            error_msg = str(e)
            self.root.after(0, lambda: self.log(f"Error starting recording: {error_msg}"))
            self.root.after(0, lambda: self.status_var.set("Error"))
            self.root.after(0, lambda: messagebox.showerror("Recording Error", f"Failed to start recording: {error_msg}"))
            self.root.after(0, lambda: self.start_button.config(state=tk.NORMAL))
    
    def stop_recording(self):
        """Stop the current recording"""
        if not self.recording or not self.recorder:
            return
        
        try:
            self.recorder.stop_recording()
            self.log("Recording stopped")
            self.recording = False
            self.recorder = None
            
            # Update UI
            self.status_var.set("Ready")
            self.start_button.config(state=tk.NORMAL)
            self.stop_button.config(state=tk.DISABLED)
            
        except Exception as e:
            self.log(f"Error stopping recording: {str(e)}")
            messagebox.showerror("Error", f"Failed to stop recording gracefully: {str(e)}")
            
            # Reset state even if there was an error
            self.recording = False
            self.recorder = None
            self.status_var.set("Ready (after error)")
            self.start_button.config(state=tk.NORMAL)
            self.stop_button.config(state=tk.DISABLED)
    
    def on_closing(self):
        """Handler for window close event"""
        if self.recording:
            if messagebox.askyesno("Quit", "Recording in progress. Stop recording and quit?"):
                self.stop_recording()
                self.root.destroy()
        else:
            self.root.destroy()

def main():
    root = tk.Tk()
    app = LSLRecorderApp(root)
    root.mainloop()

if __name__ == "__main__":
    main()
```

```python
import os
import json
import threading
import tkinter as tk
from tkinter import ttk, messagebox, scrolledtext
from glob import glob
from datetime import datetime
from pylsl import resolve_streams, StreamInfo
from liesl import Recorder

class LSLRecorderApp:
    def __init__(self, root):
        self.root = root
        self.root.title("LSL Stream Recorder")
        self.root.geometry("600x600")  # Increased size for stream list
        self.root.protocol("WM_DELETE_WINDOW", self.on_closing)
        
        # Configuration
        self.METADATA_FOLDER = r"C:\Users\cogpsy-vrlab\Documents\GitHub\BreathingSpace\Level2_RunExperiment\Results\participant_metadata"
        self.OUTPUT_FOLDER = r"C:\Users\cogpsy-vrlab\Documents\GitHub\BreathingSpace\Level2_RunExperiment\Results\LSL_Output"
        
        # Ensure output folder exists
        os.makedirs(self.OUTPUT_FOLDER, exist_ok=True)
        
        # Variables
        self.recorder = None
        self.recording = False
        self.available_streams = []
        self.selected_streams = []
        self.participant_id = tk.StringVar(value="None")
        self.status_var = tk.StringVar(value="Ready")
        self.recording_file_var = tk.StringVar(value="Not recording")
        self.streams_var = tk.StringVar(value="No streams found")
        
        # Create UI
        self.create_ui()
        
        # Try to get latest participant ID on startup
        self.refresh_participant_id()
        
        # Scan for streams on startup
        self.scan_streams()
    
    def create_ui(self):
        # Main frame
        main_frame = ttk.Frame(self.root, padding=10)
        main_frame.pack(fill=tk.BOTH, expand=True)
        
        # Title
        ttk.Label(main_frame, text="LSL Stream Recorder", font=("Arial", 16, "bold")).pack(pady=10)
        
        # Participant frame
        participant_frame = ttk.LabelFrame(main_frame, text="Participant", padding=10)
        participant_frame.pack(fill=tk.X, padx=5, pady=5)
        
        # Participant ID display
        id_frame = ttk.Frame(participant_frame)
        id_frame.pack(fill=tk.X, pady=5)
        
        ttk.Label(id_frame, text="Participant ID:").pack(side=tk.LEFT, padx=5)
        ttk.Label(id_frame, textvariable=self.participant_id, font=("Arial", 10, "bold")).pack(side=tk.LEFT, padx=5)
        ttk.Button(id_frame, text="Refresh", command=self.refresh_participant_id).pack(side=tk.RIGHT, padx=5)
        
        # Stream selection frame
        stream_frame = ttk.LabelFrame(main_frame, text="Available Streams", padding=10)
        stream_frame.pack(fill=tk.BOTH, expand=True, padx=5, pady=5)
        
        # Stream selection controls
        stream_control_frame = ttk.Frame(stream_frame)
        stream_control_frame.pack(fill=tk.X, pady=5)
        
        ttk.Button(stream_control_frame, text="Scan Streams", command=self.scan_streams).pack(side=tk.LEFT, padx=5)
        
        # Stream listbox with scrollbar
        stream_list_frame = ttk.Frame(stream_frame)
        stream_list_frame.pack(fill=tk.BOTH, expand=True)
        
        self.stream_listbox = tk.Listbox(stream_list_frame, selectmode=tk.MULTIPLE, height=10)
        self.stream_listbox.pack(side=tk.LEFT, fill=tk.BOTH, expand=True)
        
        stream_scrollbar = ttk.Scrollbar(stream_list_frame, orient=tk.VERTICAL, command=self.stream_listbox.yview)
        stream_scrollbar.pack(side=tk.RIGHT, fill=tk.Y)
        self.stream_listbox.config(yscrollcommand=stream_scrollbar.set)
        
        # Recording controls
        controls_frame = ttk.LabelFrame(main_frame, text="Recording Controls", padding=10)
        controls_frame.pack(fill=tk.X, padx=5, pady=5)
        
        # Button frame
        button_frame = ttk.Frame(controls_frame)
        button_frame.pack(fill=tk.X, pady=5)
        
        self.start_button = ttk.Button(button_frame, text="START RECORDING", command=self.start_recording, width=20)
        self.start_button.pack(side=tk.LEFT, padx=5)
        
        self.stop_button = ttk.Button(button_frame, text="STOP RECORDING", command=self.stop_recording, width=20, state=tk.DISABLED)
        self.stop_button.pack(side=tk.LEFT, padx=5)
        
        # Status frame
        status_frame = ttk.LabelFrame(main_frame, text="Status", padding=10)
        status_frame.pack(fill=tk.X, padx=5, pady=5)
        
        ttk.Label(status_frame, text="Status:").grid(row=0, column=0, sticky=tk.W, padx=5, pady=2)
        ttk.Label(status_frame, textvariable=self.status_var, font=("Arial", 10, "bold")).grid(row=0, column=1, sticky=tk.W, padx=5, pady=2)
        
        ttk.Label(status_frame, text="Recording File:").grid(row=1, column=0, sticky=tk.W, padx=5, pady=2)
        ttk.Label(status_frame, textvariable=self.recording_file_var).grid(row=1, column=1, sticky=tk.W, padx=5, pady=2)
        
        ttk.Label(status_frame, text="Streams:").grid(row=2, column=0, sticky=tk.W, padx=5, pady=2)
        ttk.Label(status_frame, textvariable=self.streams_var).grid(row=2, column=1, sticky=tk.W, padx=5, pady=2)
        
        # Log area
        log_frame = ttk.LabelFrame(main_frame, text="Log", padding=10)
        log_frame.pack(fill=tk.BOTH, expand=True, padx=5, pady=5)
        
        self.log_text = scrolledtext.ScrolledText(log_frame, height=8, wrap=tk.WORD)
        self.log_text.pack(fill=tk.BOTH, expand=True)
    
    def log(self, message):
        """Add message to log with timestamp"""
        timestamp = datetime.now().strftime("%H:%M:%S")
        log_message = f"[{timestamp}] {message}\n"
        self.log_text.insert(tk.END, log_message)
        self.log_text.see(tk.END)
        print(log_message.strip())
    
    def get_stream_info_str(self, stream_info):
        """Get a string representation of a stream for the listbox"""
        return f"{stream_info.name()} ({stream_info.type()}) - {stream_info.hostname()}"
    
    def scan_streams(self):
        """Scan for active LSL streams and update the listbox"""
        self.log("Scanning for active LSL streams...")
        self.status_var.set("Scanning streams...")
        
        # Clear listbox and streams list
        self.stream_listbox.delete(0, tk.END)
        self.available_streams = []
        
        # Scan for streams with a timeout
        try:
            # Only get active streams with a timeout
            streams = resolve_streams(wait_time=1.0)
            
            if not streams:
                self.log("No active LSL streams found")
                self.streams_var.set("No streams found")
                self.status_var.set("Ready")
                return
            
            # Add streams to listbox and store them
            for i, stream in enumerate(streams):
                try:
                    stream_str = self.get_stream_info_str(stream)
                    self.stream_listbox.insert(tk.END, stream_str)
                    self.available_streams.append(stream)
                    
                    # Log detailed stream info
                    self.log(f"Found stream: {stream.name()} (Type: {stream.type()}, Host: {stream.hostname()}, ID: {stream.source_id()})")
                except Exception as e:
                    self.log(f"Error processing stream: {str(e)}")
            
            # Update status
            self.streams_var.set(f"{len(streams)} active streams found")
            self.status_var.set("Ready")
            self.log(f"Found {len(streams)} active LSL streams")
            
        except Exception as e:
            self.log(f"Error scanning for streams: {str(e)}")
            self.status_var.set("Error scanning")
    
    def refresh_participant_id(self):
        """Get the most recent participant ID from JSON files"""
        try:
            json_files = glob(os.path.join(self.METADATA_FOLDER, "*.json"))
            
            if not json_files:
                self.log("No participant JSON files found.")
                self.participant_id.set("None")
                return
            
            # Get most recently modified file
            latest_file = max(json_files, key=os.path.getmtime)
            
            with open(latest_file, "r") as f:
                metadata = json.load(f)
            
            if "participant_id" not in metadata:
                self.log(f"participant_id missing in {os.path.basename(latest_file)}")
                self.participant_id.set("Error")
                return
            
            participant_id = metadata["participant_id"]
            self.participant_id.set(participant_id)
            self.log(f"Found participant: {participant_id}")
            
        except Exception as e:
            self.log(f"Error refreshing participant ID: {str(e)}")
            self.participant_id.set("Error")
    
    def start_recording(self):
        """Start recording selected LSL streams"""
        if self.recording:
            messagebox.showinfo("Already Recording", "Recording is already in progress.")
            return
        
        # Check if participant ID is available
        if self.participant_id.get() in ["None", "Error"]:
            messagebox.showerror("Error", "No valid participant ID found. Please refresh or check metadata folder.")
            return
        
        # Get selected streams
        selected_indices = self.stream_listbox.curselection()
        if not selected_indices:
            messagebox.showinfo("No Selection", "Please select at least one stream to record.")
            return
        
        # Store selected streams
        self.selected_streams = [self.available_streams[i] for i in selected_indices]
        
        # Start recording in a separate thread
        threading.Thread(target=self._recording_thread, daemon=True).start()
    
    def _recording_thread(self):
        """Background thread for recording to prevent UI freezing"""
        try:
            # Update UI state
            self.root.after(0, lambda: self.status_var.set("Starting recording..."))
            self.root.after(0, lambda: self.start_button.config(state=tk.DISABLED))
            
            if not self.selected_streams:
                self.root.after(0, lambda: self.log("No streams selected. Cannot start recording."))
                self.root.after(0, lambda: self.status_var.set("No streams selected"))
                self.root.after(0, lambda: self.start_button.config(state=tk.NORMAL))
                return
            
            # Log selected streams
            for stream in self.selected_streams:
                self.root.after(0, lambda s=stream: self.log(f"Will record: {s.name()} ({s.type()})"))
            
            # Update streams info
            stream_info = f"{len(self.selected_streams)} streams selected"
            self.root.after(0, lambda: self.streams_var.set(stream_info))
            
            # Prepare the output file path
            participant_id = self.participant_id.get()
            timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
            output_filename = os.path.join(self.OUTPUT_FOLDER, f"{participant_id}_{timestamp}.xdf")
            
            # Create streamargs list with necessary info
            streamargs = []
            for stream in self.selected_streams:
                # Create a dictionary with parameters to uniquely identify this stream
                stream_dict = {
                    "name": stream.name(),
                    "type": stream.type(),
                    "source_id": stream.source_id()
                }
                streamargs.append(stream_dict)
                self.root.after(0, lambda d=stream_dict: self.log(f"Adding stream: {d}"))
            
            # Create recorder and start recording
            self.recorder = Recorder()
            self.recorder.start_recording(
                filename=output_filename,
                streamargs=streamargs
            )
            
            # Update status
            self.recording = True
            self.root.after(0, lambda: self.status_var.set("Recording in progress"))
            self.root.after(0, lambda: self.recording_file_var.set(os.path.basename(output_filename)))
            self.root.after(0, lambda: self.stop_button.config(state=tk.NORMAL))
            self.root.after(0, lambda: self.log(f"Recording started to {output_filename}"))
            
        except Exception as e:
            error_msg = str(e)
            self.root.after(0, lambda: self.log(f"Error starting recording: {error_msg}"))
            self.root.after(0, lambda: self.status_var.set("Error"))
            self.root.after(0, lambda: messagebox.showerror("Recording Error", f"Failed to start recording: {error_msg}"))
            self.root.after(0, lambda: self.start_button.config(state=tk.NORMAL))
    
    def stop_recording(self):
        """Stop the current recording"""
        if not self.recording or not self.recorder:
            return
        
        try:
            self.recorder.stop_recording()
            self.log("Recording stopped")
            self.recording = False
            self.recorder = None
            
            # Update UI
            self.status_var.set("Ready")
            self.start_button.config(state=tk.NORMAL)
            self.stop_button.config(state=tk.DISABLED)
            
        except Exception as e:
            self.log(f"Error stopping recording: {str(e)}")
            messagebox.showerror("Error", f"Failed to stop recording gracefully: {str(e)}")
            
            # Reset state even if there was an error
            self.recording = False
            self.recorder = None
            self.status_var.set("Ready (after error)")
            self.start_button.config(state=tk.NORMAL)
            self.stop_button.config(state=tk.DISABLED)
    
    def on_closing(self):
        """Handler for window close event"""
        if self.recording:
            if messagebox.askyesno("Quit", "Recording in progress. Stop recording and quit?"):
                self.stop_recording()
                self.root.destroy()
        else:
            self.root.destroy()

def main():
    root = tk.Tk()
    app = LSLRecorderApp(root)
    root.mainloop()

if __name__ == "__main__":
    main()
```

```python
import os
import json
import threading
import tkinter as tk
from tkinter import ttk, messagebox
from glob import glob
from datetime import datetime
from pylsl import resolve_streams
from liesl import Recorder

class LSLRecorderApp:
    def __init__(self, root):
        self.root = root
        self.root.title("LSL Recorder")
        
        # Make window stay on top
        self.root.attributes("-topmost", True)
        
        # Get screen dimensions
        screen_width = self.root.winfo_screenwidth()
        screen_height = self.root.winfo_screenheight()
        
        # Set window size to 25% of screen
        window_width = int(screen_width * 0.25)
        window_height = int(screen_height * 0.25)
        
        # Position at bottom left
        x_position = 0
        y_position = screen_height - window_height
        
        # Set window size and position
        self.root.geometry(f"{window_width}x{window_height}+{x_position}+{y_position}")
        
        # Configuration paths
        self.METADATA_FOLDER = r"C:\Users\cogpsy-vrlab\Documents\GitHub\BreathingSpace\Level2_RunExperiment\Results\participant_metadata"
        self.OUTPUT_FOLDER = r"C:\Users\cogpsy-vrlab\Documents\GitHub\BreathingSpace\Level2_RunExperiment\Results\LSL_Output"
        
        # Ensure output folder exists
        os.makedirs(self.OUTPUT_FOLDER, exist_ok=True)
        
        # Variables
        self.recorder = None
        self.recording = False
        self.matched_streams = []
        self.participant_id = tk.StringVar(value="None")
        self.status_var = tk.StringVar(value="Initializing...")
        
        # Create UI
        self.create_ui()
        
        # Get participant ID and scan for streams
        self.refresh_participant_id()
        
        # Auto-start recording after a short delay to allow UI to initialize
        self.root.after(1000, self.auto_start_recording)
    
    def create_ui(self):
        # Main frame
        main_frame = ttk.Frame(self.root, padding=5)
        main_frame.pack(fill=tk.BOTH, expand=True)
        
        # Top section with participant info
        top_frame = ttk.Frame(main_frame)
        top_frame.pack(fill=tk.X, pady=2)
        
        # Participant ID
        ttk.Label(top_frame, text="Participant:").pack(side=tk.LEFT, padx=2)
        ttk.Label(top_frame, textvariable=self.participant_id, font=("Arial", 9, "bold")).pack(side=tk.LEFT, padx=2)
        ttk.Button(top_frame, text="Refresh", command=self.refresh_and_restart, width=7).pack(side=tk.RIGHT, padx=2)
        
        # Stream display frame
        stream_frame = ttk.LabelFrame(main_frame, text="Matched Streams (Auto-selected)", padding=3)
        stream_frame.pack(fill=tk.BOTH, expand=True, pady=2)
        
        # Stream listbox (read-only, just for display)
        self.stream_listbox = tk.Listbox(stream_frame, height=3, font=("Arial", 8), state="normal", selectmode="browse")
        self.stream_listbox.pack(side=tk.LEFT, fill=tk.BOTH, expand=True)
        
        stream_scrollbar = ttk.Scrollbar(stream_frame, orient=tk.VERTICAL, command=self.stream_listbox.yview)
        stream_scrollbar.pack(side=tk.RIGHT, fill=tk.Y)
        self.stream_listbox.config(yscrollcommand=stream_scrollbar.set)
        
        # Status and stop button
        bottom_frame = ttk.Frame(main_frame)
        bottom_frame.pack(fill=tk.X, pady=2)
        
        ttk.Label(bottom_frame, textvariable=self.status_var, font=("Arial", 8)).pack(side=tk.LEFT, padx=2)
        
        self.stop_button = ttk.Button(bottom_frame, text="STOP", command=self.stop_recording, width=5)
        self.stop_button.pack(side=tk.RIGHT, padx=2)
    
    def scan_streams(self):
        self.status_var.set("Scanning...")
        
        # Clear listbox and streams list
        self.stream_listbox.delete(0, tk.END)
        self.matched_streams = []
        
        # Get participant ID without the "P" prefix
        participant_num = self.participant_id.get().replace("P", "").strip()
        
        try:
            # Get active streams with a timeout
            all_streams = resolve_streams(wait_time=1.0)
            
            if not all_streams:
                self.status_var.set("No streams found")
                return False
            
            # Filter streams for participant ID
            for stream in all_streams:
                stream_name = stream.name().lower()
                if participant_num in stream_name or f"p{participant_num}" in stream_name:
                    self.matched_streams.append(stream)
                    self.stream_listbox.insert(tk.END, f"{stream.name()} ({stream.type()})")
            
            # If no matching streams, use all streams
            if not self.matched_streams:
                self.status_var.set("No matching streams, using all")
                self.matched_streams = all_streams
                for stream in all_streams:
                    self.stream_listbox.insert(tk.END, f"{stream.name()} ({stream.type()})")
            
            # Update status
            if self.matched_streams:
                self.status_var.set(f"{len(self.matched_streams)} streams ready")
                return True
                
        except Exception as e:
            print(f"Error scanning: {str(e)}")
            self.status_var.set("Scan error")
        
        return False
    
    def refresh_participant_id(self):
        try:
            json_files = glob(os.path.join(self.METADATA_FOLDER, "*.json"))
            
            if not json_files:
                self.participant_id.set("None")
                return False
            
            # Get most recently modified file
            latest_file = max(json_files, key=os.path.getmtime)
            
            with open(latest_file, "r") as f:
                metadata = json.load(f)
            
            if "participant_id" not in metadata:
                self.participant_id.set("Error")
                return False
            
            participant_id = metadata["participant_id"]
            self.participant_id.set(participant_id)
            print(f"Found participant: {participant_id}")
            
            # Refresh streams based on new participant
            return self.scan_streams()
            
        except Exception as e:
            print(f"Error refreshing participant ID: {str(e)}")
            self.participant_id.set("Error")
            return False
    
    def refresh_and_restart(self):
        """Refresh participant and restart recording"""
        if self.recording:
            self.stop_recording()
        
        if self.refresh_participant_id():
            self.auto_start_recording()
    
    def auto_start_recording(self):
        """Automatically start recording"""
        if self.recording:
            return
        
        if self.participant_id.get() in ["None", "Error"]:
            self.status_var.set("No valid participant ID")
            return
        
        if not self.matched_streams:
            self.status_var.set("No streams to record")
            return
        
        # Start recording in a separate thread
        threading.Thread(target=self._recording_thread, daemon=True).start()
    
    def _recording_thread(self):
        try:
            self.root.after(0, lambda: self.status_var.set("Starting..."))
            
            # Prepare output file path
            participant_id = self.participant_id.get()
            timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
            output_filename = os.path.join(self.OUTPUT_FOLDER, f"{participant_id}_{timestamp}.xdf")
            
            # Create streamargs list
            streamargs = []
            for stream in self.matched_streams:
                streamargs.append({
                    "name": stream.name(),
                    "type": stream.type(),
                    "source_id": stream.source_id()
                })
                print(f"Recording stream: {stream.name()} ({stream.type()})")
            
            # Create recorder and start recording
            self.recorder = Recorder()
            self.recorder.start_recording(filename=output_filename, streamargs=streamargs)
            
            # Update status
            self.recording = True
            self.root.after(0, lambda: self.status_var.set(f"Recording {len(self.matched_streams)} streams..."))
            print(f"Recording started to {output_filename}")
            
        except Exception as e:
            print(f"Error starting recording: {str(e)}")
            self.root.after(0, lambda: self.status_var.set("Error starting recording"))
            self.root.after(0, lambda: messagebox.showerror("Error", f"Recording error: {str(e)}"))
    
    def stop_recording(self):
        if not self.recording or not self.recorder:
            return
        
        try:
            self.recorder.stop_recording()
            print("Recording stopped")
            self.recording = False
            self.recorder = None
            
            # Update UI
            self.status_var.set("Ready - Recording stopped")
            
        except Exception as e:
            print(f"Error stopping recording: {str(e)}")
            messagebox.showerror("Error", f"Stop error: {str(e)}")
            
            # Reset state even if there was an error
            self.recording = False
            self.recorder = None
            self.status_var.set("Ready (after error)")
    
    def on_closing(self):
        if self.recording:
            if messagebox.askyesno("Quit", "Recording in progress. Stop and quit?"):
                self.stop_recording()
                self.root.destroy()
        else:
            self.root.destroy()

def main():
    root = tk.Tk()
    app = LSLRecorderApp(root)
    root.protocol("WM_DELETE_WINDOW", app.on_closing)
    root.mainloop()

if __name__ == "__main__":
    main()
```
