```python
import numpy as np
import sounddevice as sd

def generate_pulsed_sine(
    freq=50,
    sample_rate=44100,
    pulse_on_sec=0.2,
    pulse_off_sec=0.2,
    num_pulses=5,
    amplitude_db=-3
):
    """
    Generate a pulsed sine wave at the specified frequency and amplitude.
    pulse_on_sec : length of each pulse in seconds
    pulse_off_sec: silence after each pulse
    num_pulses   : how many pulses total
    amplitude_db : amplitude in dB relative to full-scale (peak=1). -3 dB => ~0.707 amplitude
    """
    # Convert amplitude in dB to a linear scale.
    amplitude = 10 ** (amplitude_db / 20.0)

    # Create one pulse of the sine
    t_on = np.linspace(0, pulse_on_sec, int(sample_rate * pulse_on_sec), endpoint=False)
    sine_on = amplitude * np.sin(2 * np.pi * freq * t_on)

    # Create the silence segment
    t_off = np.linspace(0, pulse_off_sec, int(sample_rate * pulse_off_sec), endpoint=False)
    silence_off = np.zeros_like(t_off)

    # One on+off cycle
    one_pulse = np.concatenate((sine_on, silence_off))

    # Repeat for num_pulses
    full_signal = np.tile(one_pulse, num_pulses)
    return full_signal

def main():
    # ------------- USER CONFIGURATION -------------
    # Use device indices (not names) to specify your input/output:
    # e.g., device=(5, 7) => input device #5, output device #7
    MY_DEVICE = (5, 7)     # (INPUT_INDEX, OUTPUT_INDEX)
    SAMPLE_RATE = 44100
    NUM_INPUT_CHANNELS  = 2  # e.g., "Input 3/4" is stereo
    NUM_OUTPUT_CHANNELS = 2  # sending stereo out to the Woojer
    # ----------------------------------------------

    # Generate our pulsed 50 Hz test signal at ~-3 dB
    test_signal = generate_pulsed_sine(
        freq=50,
        sample_rate=SAMPLE_RATE,
        pulse_on_sec=0.2,
        pulse_off_sec=0.2,
        num_pulses=5,
        amplitude_db=-3
    )

    # If your output is stereo, replicate the signal on both channels
    test_signal_stereo = np.column_stack((test_signal, test_signal))

    # Figure out how long the signal is
    duration_seconds = test_signal.shape[0] / float(SAMPLE_RATE)

    print("Starting simultaneous playback and recording...")
    # We'll play+record full duplex with the same sample rate, blocking until done.
    recording = sd.playrec(
        data=test_signal_stereo,
        samplerate=SAMPLE_RATE,
        channels=NUM_INPUT_CHANNELS,  # how many channels to record
        device=MY_DEVICE,             # (input_device_index, output_device_index)
        blocking=True
    )
    # 'recording' is a NumPy array of shape (num_samples, NUM_INPUT_CHANNELS)

    print("Playback and recording done. Analyzing latency...")

    # Convert recorded data to mono for the cross-correlation (use channel 0):
    recorded_mono = recording[:, 0]
    # The reference signal (mono):
    original_mono = test_signal

    # Cross-correlate to find where the recorded signal best matches the original
    corr = np.correlate(recorded_mono, original_mono, mode='full')
    # Generate the "lag" array for the correlation
    lag_arr = np.arange(-(len(original_mono)-1), len(recorded_mono))
    peak_index = np.argmax(corr)
    sample_delay = lag_arr[peak_index]

    time_delay_sec = sample_delay / float(SAMPLE_RATE)
    time_delay_millisec = time_delay_sec * 1000.0

    print(f"Estimated latency based on cross-correlation:")
    print(f"  {time_delay_sec:.5f} seconds ({time_delay_millisec:.2f} ms)")

if __name__ == "__main__":
    main()
```

```python
import time
import numpy as np
import statistics
import sounddevice as sd

##############################################################################
# 1) Ensure YOU have manually disabled "Hands-free Telephony" in
#    Control Panel → Devices and Printers → [Woojer] → Properties → Services
#    and that Windows is using the "Speakers (Woojer Strap Edge)" A2DP device.
##############################################################################

# If your version of sounddevice supports choosing a host API, 
# set default host API to WASAPI (should give more consistent latency vs. MME).
try:
    # Enumerate host APIs
    hostapis = sd.query_hostapis()
    wasapi_index = None
    for i, api in enumerate(hostapis):
        if 'Windows WASAPI' in api['name']:
            wasapi_index = i
            break
    if wasapi_index is not None:
        sd.default.hostapi = wasapi_index
        print(f"Using WASAPI host API (index {wasapi_index}).")
    else:
        print("WASAPI not found; fallback to default host API.")
except Exception as e:
    print("Could not set WASAPI host API:", e)


def generate_pulsed_sine(
    freq=50,
    sample_rate=44100,
    pulse_on_sec=0.2,
    pulse_off_sec=0.2,
    num_pulses=5,
    amplitude_db=-3
):
    """
    Generate a pulsed sine wave at the specified frequency and amplitude.
    pulse_on_sec : length of each pulse in seconds
    pulse_off_sec: silence after each pulse
    num_pulses   : how many pulses total
    amplitude_db : amplitude in dB relative to full-scale (peak=1). -3 dB => ~0.707 amplitude
    """
    amplitude = 10 ** (amplitude_db / 20.0)
    t_on = np.linspace(0, pulse_on_sec, int(sample_rate * pulse_on_sec), endpoint=False)
    sine_on = amplitude * np.sin(2 * np.pi * freq * t_on)

    t_off = np.linspace(0, pulse_off_sec, int(sample_rate * pulse_off_sec), endpoint=False)
    silence_off = np.zeros_like(t_off)

    one_pulse = np.concatenate((sine_on, silence_off))
    full_signal = np.tile(one_pulse, num_pulses)
    return full_signal


def measure_latency(
    device_pair=(5, 7),
    sample_rate=44100,
    num_input_channels=2,
    freq=50,
    amplitude_db=-3
):
    """
    Plays a pulsed freq sine wave and records simultaneously, then returns
    the measured latency in milliseconds. 
    'device_pair' must be a tuple: (input_device_index, output_device_index).
    """
    test_signal = generate_pulsed_sine(
        freq=freq,
        sample_rate=sample_rate,
        pulse_on_sec=0.2,
        pulse_off_sec=0.2,
        num_pulses=5,
        amplitude_db=amplitude_db
    )
    # If output is stereo, replicate the signal on both channels
    test_signal_stereo = np.column_stack((test_signal, test_signal))

    # Perform full-duplex play+rec
    recording = sd.playrec(
        data=test_signal_stereo,
        samplerate=sample_rate,
        channels=num_input_channels,  # how many channels to record
        device=device_pair,          # (input_index, output_index)
        blocking=True
    )

    # Convert recorded data to mono (take channel 0)
    recorded_mono = recording[:, 0]
    original_mono = test_signal

    # Cross-correlate to find offset
    corr = np.correlate(recorded_mono, original_mono, mode='full')
    lag_arr = np.arange(-(len(original_mono)-1), len(recorded_mono))
    peak_index = np.argmax(corr)
    sample_delay = lag_arr[peak_index]

    time_delay_sec = sample_delay / float(sample_rate)
    time_delay_ms = time_delay_sec * 1000.0

    return time_delay_ms


def warm_up_device(
    device_index_output=7,
    sample_rate=44100,
    duration=1.0
):
    """
    Plays 1 second of silent audio (or a short beep) to 'wake up' or
    'stabilize' the Bluetooth device in A2DP mode before measurements.
    """
    silence = np.zeros(int(sample_rate * duration))
    # Alternatively, you could do a beep:
    # beep_freq = 440
    # t = np.linspace(0, duration, int(sample_rate*duration), endpoint=False)
    # beep = 0.1 * np.sin(2*np.pi*beep_freq*t)  # low amplitude beep
    # We'll just do silence to avoid the beep in your ears:
    beep_stereo = np.column_stack((silence, silence))

    print("Playing warm-up audio (1s of silence) to stabilize device...")
    sd.play(data=beep_stereo, samplerate=sample_rate, device=device_index_output, blocking=True)
    print("Warm-up done.")


def main():
    # ---------------- USER CONFIG ----------------------
    # In your 'sd.query_devices()', find:
    #   input device index -> "Input 3/4 (Komplete Audio 6)"  (example #5)
    #   output device index -> "Speakers (Woojer Strap Edge)" (example #7)
    device_pair = (5, 7)            # (input_index, output_index)
    sample_rate = 44100
    num_input_channels = 2
    frequency = 50                  # test tone frequency
    amplitude_db = -3               # ~0.707 amplitude
    n_measurements = 5              # total number of measurement runs
    discard_first = True            # if True, discard the 1st run as a "warm-up" measurement
    # ---------------------------------------------------

    print("\n==================")
    print("Please ensure you have:")
    print(" 1) Manually disabled 'Hands-free Telephony' for Woojer")
    print(" 2) Selected 'Speakers (Woojer ...)' as the default output (A2DP)")
    print(" 3) Disabled the microphone on the Woojer if it appears in Sound devices\n")

    # (Optional) Audio warm-up to "lock" into A2DP
    warm_up_device(
        device_index_output=device_pair[1],
        sample_rate=sample_rate,
        duration=1.0  # 1 second
    )

    # Perform multiple measurements
    results = []
    for i in range(n_measurements):
        print(f"\nStarting measurement #{i+1} of {n_measurements}...")
        latency_ms = measure_latency(
            device_pair=device_pair,
            sample_rate=sample_rate,
            num_input_channels=num_input_channels,
            freq=frequency,
            amplitude_db=amplitude_db
        )
        results.append(latency_ms)
        print(f"  => Measured latency: {latency_ms:.2f} ms")

        # Optional short pause between runs
        time.sleep(1)

    # If discarding the first run (in case it’s an outlier or “settling”),
    # then remove it from the results. 
    if discard_first and len(results) > 1:
        print(f"\nDiscarding the first measurement: {results[0]:.2f} ms")
        results = results[1:]

    # Show stats
    avg_ms = statistics.mean(results) if results else 0
    stdev_ms = statistics.pstdev(results) if len(results) > 1 else 0
    min_ms = min(results) if results else 0
    max_ms = max(results) if results else 0

    print("\n=== RESULTS SUMMARY ===")
    print(f" Latency runs (ms): {results}")
    print(f" Count: {len(results)}")
    print(f" Average: {avg_ms:.2f} ms")
    print(f" StdDev : {stdev_ms:.2f} ms")
    print(f" Min    : {min_ms:.2f} ms")
    print(f" Max    : {max_ms:.2f} ms")

    print("\nAll done!\n")


if __name__ == "__main__":
    main()
```

```python
import numpy as np
import sounddevice as sd
import re
import pylsl
import time
import threading
from datetime import datetime, timedelta
import psutil
import soundfile as sf

class LSLAudioStreamer:
    def __init__(self, device_id=None, channel_name="AudioStream", 
                 sample_rate=44100, chunk_size=1024, silence_threshold=0.01, 
                 stop_after_minutes=2, shared_timer=None, audio_file=None):
        """
        Initialize an LSL audio streamer that streams audio.
        
        Parameters:
        -----------
        device_id : int
            The ID of the audio device to stream.
        channel_name : str
            The name of the LSL stream channel.
        sample_rate : int
            Audio sample rate in Hz.
        chunk_size : int
            Number of samples per chunk.
        silence_threshold : float
            The RMS threshold below which audio is considered silence.
        stop_after_minutes : int
            Minutes to continue streaming after audio is first detected.
        shared_timer : dict
            Shared timer object to synchronize multiple streamers.
        audio_file : str
            Optional audio file to stream instead of device capture
        """
        self.device_id = device_id
        self.channel_name = channel_name
        self.sample_rate = sample_rate
        self.chunk_size = chunk_size
        self.silence_threshold = silence_threshold
        self.stop_after_minutes = stop_after_minutes
        self.shared_timer = shared_timer or {'start_time': None, 'running': False, 'lock': threading.Lock()}
        self.audio_file = audio_file
        
        self.is_running = False
        
        # Get device info
        device_info = sd.query_devices(device_id) if device_id is not None else None
        self.device_name = device_info['name'] if device_info else "FileStream"
        
        # Create LSL StreamInfo and outlet
        self.info = pylsl.StreamInfo(
            name=channel_name,
            type='Audio',
            channel_count=2,  # Stereo 
            nominal_srate=sample_rate,
            channel_format='float32',
            source_id=f'audio_stream_{device_id or "file"}'
        )
        
        # Add metadata
        desc = self.info.desc()
        desc.append_child_value("device_name", self.device_name)
        
        self.outlet = pylsl.StreamOutlet(self.info)
        print(f"Created LSL outlet: {channel_name} - Will stream for {self.stop_after_minutes} minutes after audio detection")
    
    def stream_audio_file(self):
        """Stream audio from a file."""
        if not self.audio_file:
            print("No audio file specified!")
            return
        
        try:
            # Read the entire audio file
            data, file_sample_rate = sf.read(self.audio_file)
            
            # Ensure stereo
            if data.ndim == 1:
                data = np.column_stack((data, data))
            elif data.ndim > 2:
                data = data[:, :2]
            
            # Resample if necessary
            if file_sample_rate != self.sample_rate:
                from scipy import signal
                resample_ratio = self.sample_rate / file_sample_rate
                num_samples = int(np.ceil(data.shape[0] * resample_ratio))
                data = signal.resample(data, num_samples)
            
            # Normalize to float32
            data = data.astype(np.float32)
            
            # Calculate RMS for silence detection
            rms = np.sqrt(np.mean(data**2, axis=0))
            max_rms = np.max(rms)
            
            with self.shared_timer['lock']:
                print(f"Audio file loaded. Max RMS: {max_rms:.4f}")
                self.shared_timer['start_time'] = datetime.now()
                self.shared_timer['running'] = True
            
            # Stream the audio in chunks
            start_time = time.time()
            for i in range(0, len(data), self.chunk_size):
                # Check if we should stop
                with self.shared_timer['lock']:
                    if not self.is_running:
                        break
                    
                    elapsed_time = datetime.now() - self.shared_timer['start_time']
                    if elapsed_time > timedelta(minutes=self.stop_after_minutes):
                        print(f"Time limit of {self.stop_after_minutes} minutes reached. Stopping stream.")
                        self.is_running = False
                        break
                
                # Get chunk
                chunk = data[i:i+self.chunk_size]
                
                # Pad if needed
                if len(chunk) < self.chunk_size:
                    chunk = np.pad(chunk, ((0, self.chunk_size - len(chunk)), (0, 0)), mode='constant')
                
                # Push chunk to LSL
                self.outlet.push_chunk(chunk.flatten())
                
                # Control streaming rate
                time_to_wait = (i / self.sample_rate) - (time.time() - start_time)
                if time_to_wait > 0:
                    time.sleep(time_to_wait)
            
            print(f"Finished streaming audio file: {self.audio_file}")
        
        except Exception as e:
            print(f"Error streaming audio file: {e}")
        finally:
            self.is_running = False
    
    def stream_audio_device(self):
        """Stream audio from a device."""
        # This is a placeholder for device streaming if needed
        try:
            # Attempt to open device stream
            with sd.OutputStream(
                device=self.device_id,
                channels=2,
                samplerate=self.sample_rate,
                blocksize=self.chunk_size,
                callback=self.audio_callback
            ) as stream:
                while self.is_running:
                    time.sleep(0.1)
        except Exception as e:
            print(f"Error in device streaming: {e}")
        finally:
            self.is_running = False
    
    def audio_callback(self, outdata, frames, time_info, status):
        """Callback function for the audio stream."""
        if status:
            print(f"Status [{self.channel_name}]: {status}")
        
        # Calculate RMS to detect audio
        rms = np.sqrt(np.mean(outdata**2))
        
        with self.shared_timer['lock']:
            # Check if this is the first detection of audio
            if not self.shared_timer['running'] and rms > self.silence_threshold:
                print(f"Audio detected (RMS: {rms:.4f}) on device \"{self.device_name}\", starting ALL LSL streams...")
                self.shared_timer['start_time'] = datetime.now()
                self.shared_timer['running'] = True
            
            # If timer is running, check if we've exceeded the time limit
            if self.shared_timer['running']:
                elapsed_time = datetime.now() - self.shared_timer['start_time']
                if elapsed_time > timedelta(minutes=self.stop_after_minutes):
                    print(f"Time limit of {self.stop_after_minutes} minutes reached. Stopping stream [{self.channel_name}].")
                    self.is_running = False
                    return
                
                # Prepare zero output to prevent actual audio playback
                outdata.fill(0)
                
                # Send audio data to LSL - flatten stereo channels
                self.outlet.push_chunk(outdata.flatten())
    
    def start(self):
        """Start the audio streaming."""
        if self.is_running:
            print(f"Streamer [{self.channel_name}] is already running.")
            return
        
        self.is_running = True
        
        if self.audio_file:
            # Stream from audio file
            print(f"Streaming audio file: {self.audio_file}")
            self.stream_audio_file()
        else:
            # Stream from device
            print(f"Monitoring output device: ID {self.device_id}, \"{self.device_name}\" [{self.channel_name}]")
            self.stream_audio_device()
    
    def stop(self):
        """Stop the audio streaming."""
        self.is_running = False
        print(f"Stopping audio streamer [{self.channel_name}]...")


def list_audio_devices():
    """List all available audio devices with their IDs."""
    devices = sd.query_devices()
    print("\nAvailable Audio Devices:")
    print("-"*100)
    print(f"{'ID':<5} | {'Input Ch':<9} | {'Output Ch':<10} | {'Name':<70}")
    print("-"*100)
    for i, dev in enumerate(devices):
        print(f"{i:<5} | {dev['max_input_channels']:<9} | {dev['max_output_channels']:<10} | {dev['name']}")
    print("-"*100)


def find_komplete_output_devices():
    """Find Komplete Audio output devices to monitor."""
    devices = sd.query_devices()
    output1_2 = None
    output3_4 = None
    
    for i, dev in enumerate(devices):
        name_lower = dev['name'].lower()
        
        if 'komplete' in name_lower and dev['max_output_channels'] > 0:
            if 'output 1/2' in name_lower:
                output1_2 = (i, dev['name'])
            elif 'output 3/4' in name_lower:
                output3_4 = (i, dev['name'])
    
    return output1_2, output3_4


def create_clean_stream_name(prefix, device_name):
    """Create a clean LSL stream name from the device name."""
    # Remove special characters and replace spaces with underscores
    clean_name = re.sub(r'[^\w\s]', '', device_name)
    clean_name = re.sub(r'\s+', '_', clean_name)
    
    # Extract relevant parts (looking for output 1/2 or 3/4 patterns)
    output_match = re.search(r'Output_(\d+)_(\d+)', clean_name)
    if output_match:
        output_part = f"Out{output_match.group(1)}_{output_match.group(2)}"
    else:
        # Just take the first 15 chars if no specific pattern found
        output_part = clean_name[:15]
    
    # Return combined name
    return f"{prefix}_{output_part}"


def run_output_streamer(device_id, device_name, shared_timer, stop_event, audio_file=None):
    """Run a streamer for an output device in its own thread."""
    # Create clean stream name from the output device
    stream_name = create_clean_stream_name("TwoMinAudio", device_name)
    
    # Create and start the streamer with the shared timer
    streamer = LSLAudioStreamer(
        device_id=device_id,
        channel_name=stream_name,
        silence_threshold=0.01,
        shared_timer=shared_timer,
        audio_file=audio_file
    )
    
    # Start the streamer and let it run until the stop event is set
    try:
        streamer.start()
    except Exception as e:
        print(f"Error in streamer thread: {e}")
    finally:
        print(f"Streamer thread for {stream_name} exiting")


if __name__ == "__main__":
    try:
        # List available audio devices
        list_audio_devices()
        
        # Find Komplete output devices
        output1_2, output3_4 = find_komplete_output_devices()
        
        # Make sure we found the devices
        if not output1_2:
            print("Could not find Komplete Output 1/2 device! Check device names and connections.")
            exit(1)
            
        if not output3_4:
            print("Could not find Komplete Output 3/4 device! Check device names and connections.")
            exit(1)
            
        print(f"\nFound Output 1/2: ID {output1_2[0]}, {output1_2[1]}")
        print(f"Found Output 3/4: ID {output3_4[0]}, {output3_4[1]}")
        
        # Create a shared timer to synchronize both streamers
        shared_timer = {
            'start_time': None,
            'running': False,
            'lock': threading.Lock()
        }
        
        # Create a shared stop event
        stop_event = threading.Event()
        
        # Create threads for each output device
        threads = []
        
        print("\nStarting audio streamers for both output channels...")
        print("Both streams will start simultaneously when audio is detected on either channel.")
        print("Both streams will run for exactly 2 minutes from first audio detection.")
        
        # Start streamer for Output 1/2
        t1 = threading.Thread(
            target=run_output_streamer,
            args=(output1_2[0], output1_2[1], shared_timer, stop_event)
        )
        t1.daemon = True
        t1.start()
        threads.append(t1)
        
        # Start streamer for Output 3/4
        t2 = threading.Thread(
            target=run_output_streamer,
            args=(output3_4[0], output3_4[1], shared_timer, stop_event)
        )
        t2.daemon = True
        t2.start()
        threads.append(t2)
        
        print("\nBoth streamers started and synchronized. Press Ctrl+C to stop all streamers.")
        
        # Wait for all threads to finish or until interrupted
        try:
            for t in threads:
                t.join()
        except KeyboardInterrupt:
            print("\nReceived keyboard interrupt. Stopping all streamers...")
            stop_event.set()
            
            # Give threads a moment to clean up
            time.sleep(1)
            
    except Exception as e:
        print(f"\nError running script: {e}")
        import traceback
        traceback.print_exc()
        
    print("\nScript execution complete.")
import numpy as np
import sounddevice as sd
import re
import pylsl
import time
import threading
from datetime import datetime, timedelta

class LSLAudioStreamer:
    def __init__(self, device_id=None, channel_name="AudioStream", 
                 sample_rate=44100, chunk_size=1024, silence_threshold=0.01, 
                 stop_after_minutes=2, shared_timer=None):
        """
        Initialize an LSL audio streamer that streams audio from a loopback device.
        
        Parameters:
        -----------
        device_id : int
            The ID of the audio loopback device to stream.
        channel_name : str
            The name of the LSL stream channel.
        sample_rate : int
            Audio sample rate in Hz.
        chunk_size : int
            Number of samples per chunk.
        silence_threshold : float
            The RMS threshold below which audio is considered silence.
        stop_after_minutes : int
            Minutes to continue streaming after audio is first detected.
        shared_timer : dict
            Shared timer object to synchronize multiple streamers.
        """
        self.device_id = device_id
        self.channel_name = channel_name
        self.sample_rate = sample_rate
        self.chunk_size = chunk_size
        self.silence_threshold = silence_threshold
        self.stop_after_minutes = stop_after_minutes
        self.shared_timer = shared_timer or {'start_time': None, 'running': False, 'lock': threading.Lock()}
        
        self.is_running = False
        
        # Get device info
        device_info = sd.query_devices(device_id)
        self.device_name = device_info['name']
        
        # Create LSL StreamInfo and outlet
        self.info = pylsl.StreamInfo(
            name=channel_name,
            type='Audio',
            channel_count=2,  # Stereo for output devices
            nominal_srate=sample_rate,
            channel_format='float32',
            source_id=f'audio_stream_{device_id}'
        )
        
        # Add metadata
        desc = self.info.desc()
        desc.append_child_value("device_name", self.device_name)
        
        self.outlet = pylsl.StreamOutlet(self.info)
        print(f"Created LSL outlet: {channel_name} - Will stream for {self.stop_after_minutes} minutes after audio detection")
    
    def audio_callback(self, outdata, frames, time_info, status):
        """Callback function for the audio stream when monitoring outputs."""
        if status:
            print(f"Status [{self.channel_name}]: {status}")
            
        # Calculate RMS to detect audio
        rms = np.sqrt(np.mean(outdata**2))
        
        # If timer not started yet and we detect audio, start the timer
        with self.shared_timer['lock']:
            # Check if this is the first detection of audio
            if not self.shared_timer['running'] and rms > self.silence_threshold:
                print(f"Audio detected (RMS: {rms:.4f}) on device \"{self.device_name}\", starting ALL LSL streams...")
                self.shared_timer['start_time'] = datetime.now()
                self.shared_timer['running'] = True
            
            # If timer is running, check if we've exceeded the time limit
            if self.shared_timer['running']:
                elapsed_time = datetime.now() - self.shared_timer['start_time']
                if elapsed_time > timedelta(minutes=self.stop_after_minutes):
                    print(f"Time limit of {self.stop_after_minutes} minutes reached. Stopping stream [{self.channel_name}].")
                    self.is_running = False
                    return
                
                # Prepare zero output to prevent actual audio playback
                outdata.fill(0)
                
                # Send audio data to LSL - flatten stereo channels
                self.outlet.push_chunk(outdata.flatten())
                
                # Print status occasionally (only from one of the streamers to avoid spam)
                if int(elapsed_time.total_seconds()) % 15 == 0 and "1/2" in self.device_name:
                    remaining = timedelta(minutes=self.stop_after_minutes) - elapsed_time
                    print(f"Streaming audio from ALL outputs... Time remaining: {int(remaining.total_seconds())} seconds")
    
    def start(self):
        """Start the audio monitoring and LSL streaming."""
        if self.is_running:
            print(f"Streamer [{self.channel_name}] is already running.")
            return
        
        self.is_running = True
        
        print(f"Monitoring output device: ID {self.device_id}, \"{self.device_name}\" [{self.channel_name}]")
        
        # Start the audio stream
        try:
            # Use OutputStream to monitor output audio
            self.audio_stream = sd.OutputStream(
                device=self.device_id,
                channels=2,  # Stereo for output devices
                samplerate=self.sample_rate,
                blocksize=self.chunk_size,
                callback=self.audio_callback
            )
                
            self.audio_stream.start()
            
            # Keep the stream running until stopped or time limit reached
            while self.is_running:
                time.sleep(0.1)
                
        except Exception as e:
            print(f"Error [{self.channel_name}]: {e}")
        finally:
            if hasattr(self, 'audio_stream'):
                self.audio_stream.stop()
                self.audio_stream.close()
            print(f"Audio streaming stopped for [{self.channel_name}].")
    
    def stop(self):
        """Stop the audio monitoring and LSL streaming."""
        self.is_running = False
        print(f"Stopping audio streamer [{self.channel_name}]...")


def list_audio_devices():
    """List all available audio devices with their IDs."""
    devices = sd.query_devices()
    print("\nAvailable Audio Devices:")
    print("-"*100)
    print(f"{'ID':<5} | {'Input Ch':<9} | {'Output Ch':<10} | {'Name':<70}")
    print("-"*100)
    for i, dev in enumerate(devices):
        print(f"{i:<5} | {dev['max_input_channels']:<9} | {dev['max_output_channels']:<10} | {dev['name']}")
    print("-"*100)


def find_komplete_loopback_devices():
    """Find Komplete Audio output devices for direct streaming."""
    devices = sd.query_devices()
    output1_2 = None
    output3_4 = None
    
    for i, dev in enumerate(devices):
        name_lower = dev['name'].lower()
        
        # Look for Komplete Audio output devices 
        if 'komplete' in name_lower and dev['max_output_channels'] > 0:
            if 'output 1/2' in name_lower:
                output1_2 = (i, dev['name'])
            elif 'output 3/4' in name_lower:
                output3_4 = (i, dev['name'])
    
    return output1_2, output3_4


def create_clean_stream_name(prefix, device_name):
    """Create a clean LSL stream name from the device name."""
    # Remove special characters and replace spaces with underscores
    clean_name = re.sub(r'[^\w\s]', '', device_name)
    clean_name = re.sub(r'\s+', '_', clean_name)
    
    # Extract relevant parts (looking for output 1/2 or 3/4 patterns)
    output_match = re.search(r'Output_(\d+)_(\d+)', clean_name)
    if output_match:
        output_part = f"Out{output_match.group(1)}_{output_match.group(2)}"
    else:
        # Just take the first 15 chars if no specific pattern found
        output_part = clean_name[:15]
    
    # Return combined name
    return f"{prefix}_{output_part}"


def run_output_streamer(device_id, device_name, shared_timer, stop_event):
    """Run a streamer for an output device in its own thread."""
    # Create clean stream name from the output device
    stream_name = create_clean_stream_name("TwoMinAudio", device_name)
    
    # Create and start the streamer with the shared timer
    streamer = LSLAudioStreamer(
        device_id=device_id,
        channel_name=stream_name,
        silence_threshold=0.01,
        shared_timer=shared_timer
    )
    
    # Start the streamer and let it run until the stop event is set
    try:
        streamer.start()
    except Exception as e:
        print(f"Error in streamer thread: {e}")
    finally:
        print(f"Streamer thread for {stream_name} exiting")


if __name__ == "__main__":
    try:
        # List available audio devices
        list_audio_devices()
        
        # Find Komplete loopback devices
        output1_2_loopback, output3_4_loopback = find_komplete_loopback_devices()
        
        # Make sure we found the devices
        if not output1_2_loopback:
            print("Could not find Komplete Output 1/2 loopback device! Check device names, connections, and WASAPI settings.")
            exit(1)
            
        if not output3_4_loopback:
            print("Could not find Komplete Output 3/4 loopback device! Check device names, connections, and WASAPI settings.")
            exit(1)
            
        print(f"\nFound Output 1/2 Loopback: ID {output1_2_loopback[0]}, {output1_2_loopback[1]}")
        print(f"Found Output 3/4 Loopback: ID {output3_4_loopback[0]}, {output3_4_loopback[1]}")
        
        # Create a shared timer to synchronize both streamers
        shared_timer = {
            'start_time': None,
            'running': False,
            'lock': threading.Lock()
        }
        
        # Create a shared stop event
        stop_event = threading.Event()
        
        # Create threads for each output device
        threads = []
        
        print("\nStarting audio streamers for both output channels...")
        print("Both streams will start simultaneously when audio is detected on either channel.")
        print("Both streams will run for exactly 2 minutes from first audio detection.")
        
        # Start streamer for Output 1/2 Loopback
        t1 = threading.Thread(
            target=run_output_streamer,
            args=(output1_2_loopback[0], output1_2_loopback[1], shared_timer, stop_event)
        )
        t1.daemon = True
        t1.start()
        threads.append(t1)
        
        # Start streamer for Output 3/4 Loopback
        t2 = threading.Thread(
            target=run_output_streamer,
            args=(output3_4_loopback[0], output3_4_loopback[1], shared_timer, stop_event)
        )
        t2.daemon = True
        t2.start()
        threads.append(t2)
        
        print("\nBoth streamers started and synchronized. Press Ctrl+C to stop all streamers.")
        
        # Wait for all threads to finish or until interrupted
        try:
            for t in threads:
                t.join()
        except KeyboardInterrupt:
            print("\nReceived keyboard interrupt. Stopping all streamers...")
            stop_event.set()
            
            # Give threads a moment to clean up
            time.sleep(1)
            
    except Exception as e:
        print(f"\nError running script: {e}")
        import traceback
        traceback.print_exc()
        
    print("\nScript execution complete.")
```

```python
import numpy as np
import sounddevice as sd
import soundfile as sf
import pyaudio
import wave
import pylsl
import time
import threading
import queue
import sys
from datetime import datetime, timedelta
import re

class LSLAudioStreamer:
    def __init__(self, device_name=None, channel_name="AudioStream", 
                 sample_rate=44100, chunk_size=1024, silence_threshold=0.01, 
                 stop_after_minutes=2, shared_timer=None):
        """
        Initialize an LSL audio streamer using PyAudio for WASAPI loopback.
        
        Parameters:
        -----------
        device_name : str
            Name of the audio device to stream.
        channel_name : str
            The name of the LSL stream channel.
        sample_rate : int
            Audio sample rate in Hz.
        chunk_size : int
            Number of samples per chunk.
        silence_threshold : float
            The RMS threshold below which audio is considered silence.
        stop_after_minutes : int
            Minutes to continue streaming after audio is first detected.
        shared_timer : dict
            Shared timer object to synchronize multiple streamers.
        """
        self.device_name = device_name
        self.channel_name = channel_name
        self.sample_rate = sample_rate
        self.chunk_size = chunk_size
        self.silence_threshold = silence_threshold
        self.stop_after_minutes = stop_after_minutes
        
        # Use shared timer or create a new one
        self.shared_timer = shared_timer or {
            'start_time': None, 
            'running': False, 
            'lock': threading.Lock()
        }
        
        # PyAudio setup
        self.pyaudio_instance = pyaudio.PyAudio()
        self.stream = None
        self.is_running = False
        
        # Queue for audio data
        self.audio_queue = queue.Queue()
        
        # Create LSL StreamInfo and outlet
        self.info = pylsl.StreamInfo(
            name=channel_name,
            type='Audio',
            channel_count=2,  # Stereo 
            nominal_srate=sample_rate,
            channel_format='float32',
            source_id=f'audio_stream_{device_name}'
        )
        
        # Add metadata
        desc = self.info.desc()
        desc.append_child_value("device_name", device_name or "Unknown")
        
        self.outlet = pylsl.StreamOutlet(self.info)
        print(f"Created LSL outlet: {channel_name} - Will stream for {self.stop_after_minutes} minutes after audio detection")

    def find_loopback_device(self):
        """Find WASAPI loopback device index."""
        for i in range(self.pyaudio_instance.get_device_count()):
            dev_info = self.pyaudio_instance.get_device_info_by_index(i)
            if (dev_info['hostApi'] == self.pyaudio_instance.get_host_api_info_by_type(pyaudio.paWASAPI)['index'] and
                'Loopback' in dev_info['name'] and 
                self.device_name.lower() in dev_info['name'].lower()):
                return i
        return None

    def audio_callback(self, in_data, frame_count, time_info, status):
        """Callback to process audio data."""
        # Convert input data to numpy array
        audio_data = np.frombuffer(in_data, dtype=np.float32)
        
        # Reshape for stereo
        audio_data = audio_data.reshape(-1, 2)
        
        # Calculate RMS
        rms = np.sqrt(np.mean(audio_data**2))
        
        with self.shared_timer['lock']:
            # Check if this is the first detection of audio
            if not self.shared_timer['running'] and rms > self.silence_threshold:
                print(f"Audio detected (RMS: {rms:.4f}) on device \"{self.device_name}\", starting ALL LSL streams...")
                self.shared_timer['start_time'] = datetime.now()
                self.shared_timer['running'] = True
            
            # If timer is running, check if we've exceeded the time limit
            if self.shared_timer['running']:
                elapsed_time = datetime.now() - self.shared_timer['start_time']
                if elapsed_time > timedelta(minutes=self.stop_after_minutes):
                    print(f"Time limit of {self.stop_after_minutes} minutes reached. Stopping stream [{self.channel_name}].")
                    self.is_running = False
                    return (None, pyaudio.paComplete)
                
                # Push audio to LSL
                try:
                    self.outlet.push_chunk(audio_data.flatten())
                except Exception as e:
                    print(f"Error pushing to LSL: {e}")
        
        return (in_data, pyaudio.paContinue)

    def start(self):
        """Start the audio streaming."""
        if self.is_running:
            print(f"Streamer [{self.channel_name}] is already running.")
            return

        try:
            # Find the correct loopback device
            device_index = self.find_loopback_device()
            
            if device_index is None:
                print(f"Could not find loopback device for {self.device_name}")
                return

            # Open stream
            self.stream = self.pyaudio_instance.open(
                input=True,
                input_device_index=device_index,
                frames_per_buffer=self.chunk_size,
                rate=self.sample_rate,
                format=pyaudio.paFloat32,
                channels=2,
                stream_callback=self.audio_callback
            )

            self.is_running = True
            print(f"Started streaming from {self.device_name}")
            
            # Keep the thread alive
            while self.is_running:
                time.sleep(0.1)

        except Exception as e:
            print(f"Error in audio streaming: {e}")
        finally:
            # Cleanup
            if self.stream:
                self.stream.stop_stream()
                self.stream.close()
            print(f"Stopped streaming from {self.channel_name}")

    def stop(self):
        """Stop the audio streaming."""
        self.is_running = False


def list_audio_devices():
    """List all available audio devices with their IDs."""
    p = pyaudio.PyAudio()
    print("\nAvailable Audio Devices:")
    print("-"*100)
    print(f"{'ID':<5} | {'Name':<70} | {'Input Ch':<9} | {'Output Ch':<10} | {'Host API':<15}")
    print("-"*100)
    
    for i in range(p.get_device_count()):
        dev = p.get_device_info_by_index(i)
        host_api = p.get_host_api_info_by_index(dev['hostApi'])
        print(f"{i:<5} | {dev['name'][:70]:<70} | {dev['maxInputChannels']:<9} | {dev['maxOutputChannels']:<10} | {host_api['name']:<15}")
    
    print("-"*100)
    p.terminate()


def find_komplete_output_devices():
    """Find Komplete Audio output devices to monitor."""
    p = pyaudio.PyAudio()
    output1_2 = None
    output3_4 = None
    
    for i in range(p.get_device_count()):
        dev = p.get_device_info_by_index(i)
        name_lower = dev['name'].lower()
        
        if 'komplete' in name_lower and dev['maxOutputChannels'] > 0:
            if 'output 1/2' in name_lower:
                output1_2 = dev['name']
            elif 'output 3/4' in name_lower:
                output3_4 = dev['name']
    
    p.terminate()
    return output1_2, output3_4


def create_clean_stream_name(prefix, device_name):
    """Create a clean LSL stream name from the device name."""
    # Remove special characters and replace spaces with underscores
    clean_name = re.sub(r'[^\w\s]', '', device_name)
    clean_name = re.sub(r'\s+', '_', clean_name)
    
    # Extract relevant parts (looking for output 1/2 or 3/4 patterns)
    output_match = re.search(r'Output_(\d+)_(\d+)', clean_name)
    if output_match:
        output_part = f"Out{output_match.group(1)}_{output_match.group(2)}"
    else:
        # Just take the first 15 chars if no specific pattern found
        output_part = clean_name[:15]
    
    # Return combined name
    return f"{prefix}_{output_part}"


def run_output_streamer(device_name, shared_timer, stop_event):
    """Run a streamer for an output device in its own thread."""
    # Create clean stream name from the output device
    stream_name = create_clean_stream_name("TwoMinAudio", device_name)
    
    # Create and start the streamer with the shared timer
    streamer = LSLAudioStreamer(
        device_name=device_name,
        channel_name=stream_name,
        silence_threshold=0.01,
        shared_timer=shared_timer
    )
    
    # Start the streamer and let it run until the stop event is set
    try:
        streamer.start()
    except Exception as e:
        print(f"Error in streamer thread: {e}")
    finally:
        print(f"Streamer thread for {stream_name} exiting")


if __name__ == "__main__":
    # Ensure required libraries are imported
    try:
        import pyaudio
    except ImportError:
        print("Please install PyAudio: pip install pyaudio")
        sys.exit(1)

    try:
        # List available audio devices
        list_audio_devices()
        
        # Find Komplete output devices
        output1_2, output3_4 = find_komplete_output_devices()
        
        # Make sure we found the devices
        if not output1_2:
            print("Could not find Komplete Output 1/2 device! Check device names and connections.")
            exit(1)
            
        if not output3_4:
            print("Could not find Komplete Output 3/4 device! Check device names and connections.")
            exit(1)
            
        print(f"\nFound Output 1/2: {output1_2}")
        print(f"Found Output 3/4: {output3_4}")
        
        # Create a shared timer to synchronize both streamers
        shared_timer = {
            'start_time': None,
            'running': False,
            'lock': threading.Lock()
        }
        
        # Create a shared stop event
        stop_event = threading.Event()
        
        # Create threads for each output device
        threads = []
        
        print("\nStarting audio streamers for both output channels...")
        print("Both streams will start simultaneously when audio is detected on either channel.")
        print("Both streams will run for exactly 2 minutes from first audio detection.")
        
        # Start streamer for Output 1/2
        t1 = threading.Thread(
            target=run_output_streamer,
            args=(output1_2, shared_timer, stop_event)
        )
        t1.daemon = True
        t1.start()
        threads.append(t1)
        
        # Start streamer for Output 3/4
        t2 = threading.Thread(
            target=run_output_streamer,
            args=(output3_4, shared_timer, stop_event)
        )
        t2.daemon = True
        t2.start()
        threads.append(t2)
        
        print("\nBoth streamers started and synchronized. Press Ctrl+C to stop all streamers.")
        
        # Wait for all threads to finish or until interrupted
        try:
            for t in threads:
                t.join()
        except KeyboardInterrupt:
            print("\nReceived keyboard interrupt. Stopping all streamers...")
            stop_event.set()
            
            # Give threads a moment to clean up
            time.sleep(1)
            
    except Exception as e:
        print(f"\nError running script: {e}")
        import traceback
        traceback.print_exc()
        
    print("\nScript execution complete.")
```

```python
import numpy as np
import pyaudio
import pylsl
import time
import threading
import re
from datetime import datetime, timedelta
import sys

def list_audio_devices():
    """List all available audio devices with their IDs."""
    p = pyaudio.PyAudio()
    print("\nAvailable Audio Devices:")
    print("-"*100)
    print(f"{'ID':<5} | {'Name':<70} | {'Input Ch':<9} | {'Output Ch':<10} | {'Host API':<15}")
    print("-"*100)
    
    try:
        for i in range(p.get_device_count()):
            try:
                dev = p.get_device_info_by_index(i)
                host_api = p.get_host_api_info_by_index(dev['hostApi'])
                print(f"{i:<5} | {dev['name'][:70]:<70} | {dev['maxInputChannels']:<9} | {dev['maxOutputChannels']:<10} | {host_api['name']:<15}")
            except Exception as dev_err:
                print(f"{i:<5} | Error retrieving device info: {dev_err}")
    except Exception as e:
        print(f"Error listing devices: {e}")
    finally:
        p.terminate()
    
    print("-"*100)

def find_komplete_output_devices():
    """Find Komplete Audio output devices to monitor."""
    p = pyaudio.PyAudio()
    output1_2 = None
    output3_4 = None
    
    try:
        for i in range(p.get_device_count()):
            try:
                dev = p.get_device_info_by_index(i)
                name_lower = dev['name'].lower()
                
                # Look for output devices with Komplete in the name
                if dev['maxOutputChannels'] > 0 and 'komplete' in name_lower:
                    if 'output 1/2' in name_lower:
                        output1_2 = (i, dev['name'], dev['maxOutputChannels'])
                    elif 'output 3/4' in name_lower:
                        output3_4 = (i, dev['name'], dev['maxOutputChannels'])
            except Exception:
                # Skip devices that can't be read
                continue
    finally:
        p.terminate()
    
    return output1_2, output3_4

def create_clean_stream_name(prefix, device_name):
    """Create a clean LSL stream name from the device name."""
    # Remove special characters and replace spaces with underscores
    clean_name = re.sub(r'[^\w\s]', '', device_name)
    clean_name = re.sub(r'\s+', '_', clean_name)
    
    # Extract relevant parts (looking for output 1/2 or 3/4 patterns)
    output_match = re.search(r'Output_(\d+)_(\d+)', clean_name)
    if output_match:
        output_part = f"Out{output_match.group(1)}_{output_match.group(2)}"
    else:
        # Just take the first 15 chars if no specific pattern found
        output_part = clean_name[:15]
    
    # Return combined name
    return f"{prefix}_{output_part}"

class LSLAudioStreamer:
    def __init__(self, device_id=None, device_name=None, channel_name="AudioStream", 
                 sample_rate=44100, chunk_size=1024, silence_threshold=0.01, 
                 stop_after_minutes=2, shared_timer=None, num_channels=2):
        """
        Initialize an LSL audio streamer.
        
        Parameters:
        -----------
        device_id : int
            Device index to stream from.
        device_name : str
            Name of the audio device.
        channel_name : str
            The name of the LSL stream channel.
        sample_rate : int
            Audio sample rate in Hz.
        chunk_size : int
            Number of samples per chunk.
        silence_threshold : float
            The RMS threshold below which audio is considered silence.
        stop_after_minutes : int
            Minutes to continue streaming after audio is first detected.
        shared_timer : dict
            Shared timer object to synchronize multiple streamers.
        num_channels : int
            Number of audio channels to use.
        """
        self.device_id = device_id
        self.device_name = device_name
        self.channel_name = channel_name
        self.sample_rate = sample_rate
        self.chunk_size = chunk_size
        self.silence_threshold = silence_threshold
        self.stop_after_minutes = stop_after_minutes
        self.num_channels = num_channels
        
        # Use shared timer or create a new one
        self.shared_timer = shared_timer or {
            'start_time': None, 
            'running': False, 
            'lock': threading.Lock()
        }
        
        # PyAudio setup
        self.pyaudio_instance = pyaudio.PyAudio()
        self.stream = None
        self.is_running = False
        
        # Verify device info safely
        try:
            # We'll use p.get_device_info_by_index() with error handling
            def safe_get_device_info(index):
                try:
                    return self.pyaudio_instance.get_device_info_by_index(index)
                except Exception as e:
                    print(f"Error getting device info for index {index}: {e}")
                    return None
            
            dev_info = safe_get_device_info(device_id)
            if dev_info:
                print(f"Streaming from device: {dev_info.get('name', 'Unknown')}")
                print(f"Max Channels: {dev_info.get('maxOutputChannels', 'N/A')}")
            else:
                print(f"Could not retrieve device info for index {device_id}")
        except Exception as e:
            print(f"Error in device info retrieval: {e}")
        
        # Create LSL StreamInfo and outlet
        self.info = pylsl.StreamInfo(
            name=channel_name,
            type='Audio',
            channel_count=self.num_channels,  # Use specified number of channels
            nominal_srate=sample_rate,
            channel_format='float32',
            source_id=f'audio_stream_{device_id}'
        )
        
        # Add metadata
        desc = self.info.desc()
        desc.append_child_value("device_name", device_name or "Unknown")
        
        self.outlet = pylsl.StreamOutlet(self.info)
        print(f"Created LSL outlet: {channel_name} - Will stream for {self.stop_after_minutes} minutes after audio detection")

    def audio_callback(self, in_data, frame_count, time_info, status):
        """Callback to process audio data."""
        # Convert input data to numpy array
        try:
            audio_data = np.frombuffer(in_data, dtype=np.float32)
            
            # Reshape for specified number of channels
            audio_data = audio_data.reshape(-1, self.num_channels)
            
            # Calculate RMS
            rms = np.sqrt(np.mean(audio_data**2))
            
            with self.shared_timer['lock']:
                # Check if this is the first detection of audio
                if not self.shared_timer['running'] and rms > self.silence_threshold:
                    print(f"Audio detected (RMS: {rms:.4f}) on device \"{self.device_name}\", starting ALL LSL streams...")
                    self.shared_timer['start_time'] = datetime.now()
                    self.shared_timer['running'] = True
                
                # If timer is running, check if we've exceeded the time limit
                if self.shared_timer['running']:
                    elapsed_time = datetime.now() - self.shared_timer['start_time']
                    if elapsed_time > timedelta(minutes=self.stop_after_minutes):
                        print(f"Time limit of {self.stop_after_minutes} minutes reached. Stopping stream [{self.channel_name}].")
                        self.is_running = False
                        return (None, pyaudio.paComplete)
                    
                    # Push audio to LSL
                    try:
                        self.outlet.push_chunk(audio_data.flatten())
                    except Exception as e:
                        print(f"Error pushing to LSL: {e}")
            
            return (in_data, pyaudio.paContinue)
        
        except Exception as e:
            print(f"Error in audio callback: {e}")
            return (None, pyaudio.paComplete)

    def start(self):
        """Start the audio streaming."""
        if self.is_running:
            print(f"Streamer [{self.channel_name}] is already running.")
            return

        try:
            # Open stream
            self.stream = self.pyaudio_instance.open(
                input=True,
                input_device_index=self.device_id,
                frames_per_buffer=self.chunk_size,
                rate=self.sample_rate,
                format=pyaudio.paFloat32,
                channels=self.num_channels,
                stream_callback=self.audio_callback
            )

            self.is_running = True
            print(f"Started streaming from {self.device_name}")
            
            # Keep the thread alive
            while self.is_running:
                time.sleep(0.1)

        except Exception as e:
            print(f"Error in audio streaming: {e}")
        finally:
            # Cleanup
            if self.stream:
                self.stream.stop_stream()
                self.stream.close()
            print(f"Stopped streaming from {self.channel_name}")

    def stop(self):
        """Stop the audio streaming."""
        self.is_running = False
        if self.stream:
            self.stream.stop_stream()
        self.pyaudio_instance.terminate()

def run_output_streamer(device_info, shared_timer, stop_event):
    """Run a streamer for an output device in its own thread."""
    # Unpack device info
    device_id, device_name, max_channels = device_info
    
    # Create clean stream name from the output device
    stream_name = create_clean_stream_name("TwoMinAudio", device_name)
    
    # Determine number of channels (cap at 2)
    num_channels = min(max_channels, 2)
    
    # Create and start the streamer with the shared timer
    streamer = LSLAudioStreamer(
        device_id=device_id,
        device_name=device_name,
        channel_name=stream_name,
        silence_threshold=0.01,
        shared_timer=shared_timer,
        num_channels=num_channels
    )
    
    # Start the streamer and let it run until the stop event is set
    try:
        streamer.start()
    except Exception as e:
        print(f"Error in streamer thread: {e}")
    finally:
        print(f"Streamer thread for {stream_name} exiting")

def main():
    try:
        # List available audio devices
        list_audio_devices()
        
        # Find Komplete output devices
        output1_2, output3_4 = find_komplete_output_devices()
        
        # Make sure we found the devices
        if not output1_2:
            print("Could not find Komplete Output 1/2 device! Check device names and connections.")
            return
            
        if not output3_4:
            print("Could not find Komplete Output 3/4 device! Check device names and connections.")
            return
            
        print(f"\nFound Output 1/2: ID {output1_2[0]}, {output1_2[1]}, Channels: {output1_2[2]}")
        print(f"Found Output 3/4: ID {output3_4[0]}, {output3_4[1]}, Channels: {output3_4[2]}")
        
        # Create a shared timer to synchronize both streamers
        shared_timer = {
            'start_time': None,
            'running': False,
            'lock': threading.Lock()
        }
        
        # Create a shared stop event
        stop_event = threading.Event()
        
        # Create threads for each output device
        threads = []
        
        print("\nStarting audio streamers for both output channels...")
        print("Both streams will start simultaneously when audio is detected on either channel.")
        print("Both streams will run for exactly 2 minutes from first audio detection.")
        
        # Start streamer for Output 1/2
        t1 = threading.Thread(
            target=run_output_streamer,
            args=(output1_2, shared_timer, stop_event)
        )
        t1.daemon = True
        t1.start()
        threads.append(t1)
        
        # Start streamer for Output 3/4
        t2 = threading.Thread(
            target=run_output_streamer,
            args=(output3_4, shared_timer, stop_event)
        )
        t2.daemon = True
        t2.start()
        threads.append(t2)
        
        print("\nBoth streamers started and synchronized. Press Ctrl+C to stop all streamers.")
        
        # Wait for all threads to finish or until interrupted
        try:
            for t in threads:
                t.join()
        except KeyboardInterrupt:
            print("\nReceived keyboard interrupt. Stopping all streamers...")
            stop_event.set()
            
            # Give threads a moment to clean up
            time.sleep(1)
            
    except Exception as e:
        print(f"\nError running script: {e}")
        import traceback
        traceback.print_exc()
        
    print("\nScript execution complete.")

# For Jupyter notebook compatibility
if __name__ == "__main__" or __name__ == "__main__\n":
    # Check if running in a Jupyter notebook
    try:
        get_ipython
        main()
    except NameError:
        # Not in a Jupyter notebook, run normally
        main()
```

```python
import numpy as np
import pyaudio
import pylsl
import time
import threading
import re
from datetime import datetime, timedelta
import sys
import traceback

# Global debug flag - set to True for verbose logging
DEBUG = True

def debug_log(message):
    """Print debug messages with timestamp if DEBUG is True."""
    if DEBUG:
        timestamp = datetime.now().strftime("%H:%M:%S.%f")[:-3]
        print(f"[DEBUG {timestamp}] {message}")

def list_audio_devices():
    """List all available audio devices with their IDs."""
    debug_log("Listing audio devices...")
    p = pyaudio.PyAudio()
    print("\nAvailable Audio Devices:")
    print("-"*100)
    print(f"{'ID':<5} | {'Name':<70} | {'Input Ch':<9} | {'Output Ch':<10} | {'Host API':<15}")
    print("-"*100)
    
    try:
        for i in range(p.get_device_count()):
            try:
                dev = p.get_device_info_by_index(i)
                host_api = p.get_host_api_info_by_index(dev['hostApi'])
                print(f"{i:<5} | {dev['name'][:70]:<70} | {dev['maxInputChannels']:<9} | {dev['maxOutputChannels']:<10} | {host_api['name']:<15}")
            except Exception as dev_err:
                print(f"{i:<5} | Error retrieving device info: {dev_err}")
    except Exception as e:
        print(f"Error listing devices: {e}")
    finally:
        p.terminate()
    
    print("-"*100)
    debug_log("Finished listing audio devices")

def find_komplete_output_devices():
    """Find Komplete Audio output devices to monitor."""
    debug_log("Searching for Komplete Audio output devices...")
    p = pyaudio.PyAudio()
    output1_2 = None
    output3_4 = None
    
    try:
        for i in range(p.get_device_count()):
            try:
                dev = p.get_device_info_by_index(i)
                name_lower = dev['name'].lower()
                
                # Look for Komplete output devices across different host APIs
                if dev['maxOutputChannels'] > 0 and 'komplete' in name_lower:
                    debug_log(f"Found potential Komplete device: {i} - {dev['name']}")
                    
                    # Prefer WASAPI or devices with definitive output configuration
                    if ('output 1/2' in name_lower or 'output 1 2' in name_lower):
                        host_api = p.get_host_api_info_by_index(dev['hostApi'])['name']
                        debug_log(f"  - Matches Output 1/2 pattern, Host API: {host_api}")
                        
                        if (output1_2 is None or 'wasapi' in host_api.lower()):
                            output1_2 = (i, dev['name'], dev['maxOutputChannels'])
                            debug_log(f"  - Selected as Output 1/2: {output1_2}")
                    
                    elif ('output 3/4' in name_lower or 'output 3 4' in name_lower):
                        host_api = p.get_host_api_info_by_index(dev['hostApi'])['name']
                        debug_log(f"  - Matches Output 3/4 pattern, Host API: {host_api}")
                        
                        if (output3_4 is None or 'wasapi' in host_api.lower()):
                            output3_4 = (i, dev['name'], dev['maxOutputChannels'])
                            debug_log(f"  - Selected as Output 3/4: {output3_4}")
            except Exception as e:
                debug_log(f"Error checking device {i}: {e}")
                # Skip devices that can't be read
                continue
    finally:
        p.terminate()
    
    debug_log(f"Found Output 1/2: {output1_2}")
    debug_log(f"Found Output 3/4: {output3_4}")
    
    return output1_2, output3_4

def create_clean_stream_name(prefix, device_name):
    """Create a clean LSL stream name from the device name."""
    debug_log(f"Creating clean stream name from: {device_name}")
    
    # Remove special characters and replace spaces with underscores
    clean_name = re.sub(r'[^\w\s]', '', device_name)
    clean_name = re.sub(r'\s+', '_', clean_name)
    
    # Extract relevant parts (looking for output 1/2 or 3/4 patterns)
    output_match = re.search(r'Output_(\d+)_(\d+)', clean_name)
    if output_match:
        output_part = f"Out{output_match.group(1)}_{output_match.group(2)}"
    else:
        # Just take the first 15 chars if no specific pattern found
        output_part = clean_name[:15]
    
    # Return combined name
    result = f"{prefix}_{output_part}"
    debug_log(f"Clean stream name: {result}")
    return result

class LSLAudioStreamer:
    def __init__(self, device_id=None, device_name=None, channel_name="AudioStream", 
                 sample_rate=44100, chunk_size=1024, silence_threshold=0.01, 
                 stop_after_minutes=2, shared_timer=None, num_channels=2, 
                 is_input_device=False):
        """
        Initialize an LSL audio streamer.
        
        Parameters:
        -----------
        device_id : int
            Device index to stream from.
        device_name : str
            Name of the audio device.
        channel_name : str
            The name of the LSL stream channel.
        sample_rate : int
            Audio sample rate in Hz.
        chunk_size : int
            Number of samples per chunk.
        silence_threshold : float
            The RMS threshold below which audio is considered silence.
        stop_after_minutes : int
            Minutes to continue streaming after audio is first detected.
        shared_timer : dict
            Shared timer object to synchronize multiple streamers.
        num_channels : int
            Number of audio channels to use.
        is_input_device : bool
            Whether this is an input device (True) or output device (False)
        """
        debug_log(f"Initializing LSLAudioStreamer for device {device_id} ({device_name})")
        self.device_id = device_id
        self.device_name = device_name
        self.channel_name = channel_name
        self.sample_rate = sample_rate
        self.chunk_size = chunk_size
        self.silence_threshold = silence_threshold
        self.stop_after_minutes = stop_after_minutes
        self.num_channels = num_channels
        self.is_input_device = is_input_device
        
        # Use shared timer or create a new one
        self.shared_timer = shared_timer or {
            'start_time': None, 
            'running': False, 
            'lock': threading.Lock()
        }
        
        # PyAudio setup
        self.pyaudio_instance = pyaudio.PyAudio()
        self.stream = None
        self.is_running = False
        
        # Verify device info safely
        try:
            # We'll use p.get_device_info_by_index() with error handling
            def safe_get_device_info(index):
                try:
                    return self.pyaudio_instance.get_device_info_by_index(index)
                except Exception as e:
                    debug_log(f"Error getting device info for index {index}: {e}")
                    return None
            
            dev_info = safe_get_device_info(device_id)
            if dev_info:
                debug_log(f"Device information for ID {device_id}:")
                for key, value in dev_info.items():
                    debug_log(f"  {key}: {value}")
                    
                print(f"Streaming from device: {dev_info.get('name', 'Unknown')}")
                if self.is_input_device:
                    print(f"Max Input Channels: {dev_info.get('maxInputChannels', 'N/A')}")
                else:
                    print(f"Max Output Channels: {dev_info.get('maxOutputChannels', 'N/A')}")
            else:
                print(f"Could not retrieve device info for index {device_id}")
        except Exception as e:
            print(f"Error in device info retrieval: {e}")
            debug_log(f"Exception: {traceback.format_exc()}")
        
        # Create LSL StreamInfo and outlet
        try:
            debug_log(f"Creating LSL StreamInfo with {self.num_channels} channels")
            self.info = pylsl.StreamInfo(
                name=channel_name,
                type='Audio',
                channel_count=self.num_channels,
                nominal_srate=sample_rate,
                channel_format='float32',
                source_id=f'audio_stream_{device_id}'
            )
            
            # Add metadata
            desc = self.info.desc()
            desc.append_child_value("device_name", device_name or "Unknown")
            
            debug_log("Creating LSL outlet")
            self.outlet = pylsl.StreamOutlet(self.info)
            print(f"Created LSL outlet: {channel_name} - Will stream for {self.stop_after_minutes} minutes after audio detection")
        except Exception as e:
            print(f"Error creating LSL outlet: {e}")
            debug_log(f"Exception: {traceback.format_exc()}")

    def audio_callback(self, in_data, frame_count, time_info, status):
        """Callback to process audio data."""
        # Convert input data to numpy array
        try:
            if status:
                debug_log(f"Audio callback status: {status}")
                
            audio_data = np.frombuffer(in_data, dtype=np.float32)
            
            # Reshape for specified number of channels
            audio_data = audio_data.reshape(-1, self.num_channels)
            
            # Calculate RMS
            rms = np.sqrt(np.mean(audio_data**2))
            
            # Log RMS values periodically to see audio activity
            if hasattr(self, 'callback_counter'):
                self.callback_counter += 1
                if self.callback_counter % 100 == 0:  # Log every 100 calls
                    debug_log(f"Audio callback RMS level: {rms:.6f}")
            else:
                self.callback_counter = 0
            
            with self.shared_timer['lock']:
                # Check if this is the first detection of audio
                if not self.shared_timer['running'] and rms > self.silence_threshold:
                    print(f"Audio detected (RMS: {rms:.4f}) on device \"{self.device_name}\", starting ALL LSL streams...")
                    self.shared_timer['start_time'] = datetime.now()
                    self.shared_timer['running'] = True
                    debug_log(f"Audio detected above threshold {self.silence_threshold}. Starting timer.")
                
                # If timer is running, check if we've exceeded the time limit
                if self.shared_timer['running']:
                    elapsed_time = datetime.now() - self.shared_timer['start_time']
                    if elapsed_time > timedelta(minutes=self.stop_after_minutes):
                        print(f"Time limit of {self.stop_after_minutes} minutes reached. Stopping stream [{self.channel_name}].")
                        debug_log("Time limit reached, stopping stream")
                        self.is_running = False
                        return (None, pyaudio.paComplete)
                    
                    # Push audio to LSL
                    try:
                        self.outlet.push_chunk(audio_data.flatten())
                    except Exception as e:
                        print(f"Error pushing to LSL: {e}")
                        debug_log(f"LSL push error: {e}")
            
            return (in_data, pyaudio.paContinue)
        
        except Exception as e:
            print(f"Error in audio callback: {e}")
            debug_log(f"Exception in audio callback: {traceback.format_exc()}")
            return (None, pyaudio.paComplete)

    def start(self):
        """Start the audio streaming."""
        debug_log(f"Starting audio streamer for {self.device_name} (ID: {self.device_id})")
        if self.is_running:
            print(f"Streamer [{self.channel_name}] is already running.")
            return

        try:
            # For monitoring output devices, we need to use loopback mode
            # This is only available in some APIs (like WASAPI)
            if self.is_input_device:
                # Open as input device
                debug_log(f"Opening {self.device_name} as INPUT device")
                print(f"Opening {self.device_name} as INPUT device")
                self.stream = self.pyaudio_instance.open(
                    input=True,
                    input_device_index=self.device_id,
                    frames_per_buffer=self.chunk_size,
                    rate=self.sample_rate,
                    format=pyaudio.paFloat32,
                    channels=self.num_channels,
                    stream_callback=self.audio_callback
                )
                debug_log("Successfully opened input stream")
            else:
                # For output devices, try to use WASAPI loopback mode
                p = self.pyaudio_instance
                
                # Get device info to check host API
                try:
                    dev_info = p.get_device_info_by_index(self.device_id)
                    host_api = p.get_host_api_info_by_index(dev_info["hostApi"])
                    host_api_name = host_api.get("name", "")
                    
                    debug_log(f"Device host API: {host_api_name}")
                    is_wasapi = "wasapi" in host_api_name.lower()
                    print(f"Device uses {host_api_name}")
                    
                    if is_wasapi:
                        # WASAPI can do loopback (capture what's being played)
                        debug_log(f"Opening {self.device_name} with WASAPI loopback")
                        print(f"Opening {self.device_name} with WASAPI loopback")
                        try:
                            self.stream = p.open(
                                format=pyaudio.paFloat32,
                                channels=self.num_channels,
                                rate=self.sample_rate,
                                frames_per_buffer=self.chunk_size,
                                input=True,
                                input_device_index=self.device_id,
                                stream_callback=self.audio_callback,
                                # This is the key for WASAPI loopback
                                as_loopback=True
                            )
                            debug_log("Successfully opened WASAPI loopback stream")
                        except Exception as e:
                            debug_log(f"Error opening WASAPI loopback: {e}")
                            debug_log(f"Exception: {traceback.format_exc()}")
                            raise
                    else:
                        # For non-WASAPI, we need to find corresponding input
                        debug_log(f"Non-WASAPI output device detected. Searching for corresponding input...")
                        print(f"Non-WASAPI output device detected. Searching for corresponding input...")
                        
                        # Search for a related input device with the same name pattern
                        related_input_id = None
                        for i in range(p.get_device_count()):
                            try:
                                info = p.get_device_info_by_index(i)
                                # Check if it's an input device with a similar name
                                if (info["maxInputChannels"] > 0 and 
                                    self.device_name.split("(")[-1] in info["name"]):
                                    related_input_id = i
                                    debug_log(f"Found related input device: {i} - {info['name']}")
                                    print(f"Found related input device: {info['name']}")
                                    break
                            except Exception as e:
                                debug_log(f"Error checking device {i}: {e}")
                                continue
                                
                        if related_input_id is not None:
                            debug_log(f"Opening related input device: {related_input_id}")
                            print(f"Opening related input device: {related_input_id}")
                            self.stream = p.open(
                                format=pyaudio.paFloat32,
                                channels=self.num_channels,
                                rate=self.sample_rate,
                                frames_per_buffer=self.chunk_size,
                                input=True,
                                input_device_index=related_input_id,
                                stream_callback=self.audio_callback
                            )
                            debug_log("Successfully opened related input stream")
                        else:
                            # If we can't find a related input, try opening a dummy output
                            # (This won't capture audio, but might avoid errors)
                            debug_log(f"Warning: No related input found. Using silent output stream.")
                            print(f"Warning: No related input found. Using silent output stream.")
                            
                            # Open output device
                            self.stream = p.open(
                                format=pyaudio.paFloat32,
                                channels=self.num_channels,
                                rate=self.sample_rate,
                                frames_per_buffer=self.chunk_size,
                                output=True,
                                output_device_index=self.device_id
                            )
                            debug_log("Successfully opened silent output stream")
                            # The problem is we can't monitor this output without an input
                            print("WARNING: Can't monitor this output device!")
                except Exception as e:
                    debug_log(f"Error inspecting device: {e}")
                    debug_log(f"Exception: {traceback.format_exc()}")
                    print(f"Error inspecting device: {e}")
                    
                    # Fallback - try opening as input device (will likely fail)
                    debug_log("Fallback: attempting to open as input device (may fail)")
                    print("Fallback: attempting to open as input device (may fail)")
                    try:
                        self.stream = p.open(
                            input=True,
                            input_device_index=self.device_id,
                            frames_per_buffer=self.chunk_size,
                            rate=self.sample_rate,
                            format=pyaudio.paFloat32,
                            channels=self.num_channels,
                            stream_callback=self.audio_callback
                        )
                        debug_log("Successfully opened fallback input stream")
                    except Exception as fallback_e:
                        debug_log(f"Fallback also failed: {fallback_e}")
                        debug_log(f"Exception: {traceback.format_exc()}")
                        raise

            self.is_running = True
            print(f"Started streaming from {self.device_name}")
            debug_log(f"Successfully started streaming from {self.device_name}")
            
            # Keep the thread alive
            debug_log("Entering main stream loop")
            counter = 0
            while self.is_running:
                time.sleep(0.1)
                counter += 1
                if counter % 100 == 0:  # Log every 10 seconds
                    debug_log(f"Stream for {self.channel_name} still running...")

        except Exception as e:
            print(f"Error in audio streaming: {e}")
            debug_log(f"Error in audio streaming: {e}")
            debug_log(f"Exception: {traceback.format_exc()}")
        finally:
            # Cleanup
            debug_log("Cleaning up stream resources")
            if self.stream:
                try:
                    self.stream.stop_stream()
                    self.stream.close()
                except Exception as e:
                    debug_log(f"Error closing stream: {e}")
            print(f"Stopped streaming from {self.channel_name}")
            debug_log(f"Stopped streaming from {self.channel_name}")

    def stop(self):
        """Stop the audio streaming."""
        debug_log(f"Stopping stream for {self.channel_name}")
        self.is_running = False
        if self.stream:
            try:
                self.stream.stop_stream()
            except Exception as e:
                debug_log(f"Error stopping stream: {e}")
        try:
            self.pyaudio_instance.terminate()
            debug_log("PyAudio instance terminated")
        except Exception as e:
            debug_log(f"Error terminating PyAudio: {e}")

def run_output_streamer(device_info, shared_timer, stop_event):
    """Run a streamer for an output device in its own thread."""
    debug_log(f"Starting output streamer thread for device: {device_info}")
    
    # Unpack device info
    device_id, device_name, max_channels = device_info
    
    # Create clean stream name from the output device
    stream_name = create_clean_stream_name("TwoMinAudio", device_name)
    
    # Diagnostic print
    print(f"\nAttempting to stream from device: {device_name}")
    print(f"Device ID: {device_id}")
    
    # Try to get more detailed device information
    p = pyaudio.PyAudio()
    try:
        dev_info = p.get_device_info_by_index(device_id)
        print("Detailed Device Info:")
        print(f"  Name: {dev_info.get('name', 'N/A')}")
        print(f"  Max Input Channels: {dev_info.get('maxInputChannels', 'N/A')}")
        print(f"  Max Output Channels: {dev_info.get('maxOutputChannels', 'N/A')}")
        print(f"  Default Sample Rate: {dev_info.get('defaultSampleRate', 'N/A')}")
        host_api = p.get_host_api_info_by_index(dev_info['hostApi'])['name']
        print(f"  Host API: {host_api}")
        
        # Determine if we have WASAPI loopback capability
        has_loopback = "wasapi" in host_api.lower()
        if has_loopback:
            print("  WASAPI loopback mode available")
        else:
            print("  WASAPI loopback not available - may need alternative approach")
    except Exception as e:
        print(f"Error retrieving detailed device info: {e}")
        debug_log(f"Exception: {traceback.format_exc()}")
    finally:
        p.terminate()
    
    # Determine appropriate channel configuration
    # First try a mix of input and output channels
    p = pyaudio.PyAudio()
    try:
        # Get full device info
        dev_info = p.get_device_info_by_index(device_id)
        
        # Determine number of channels to use
        input_channels = dev_info.get('maxInputChannels', 0)
        output_channels = dev_info.get('maxOutputChannels', 0)
        
        # Determine if this is an input or output device
        is_input_device = input_channels > 0 and input_channels >= output_channels
        
        # Choose appropriate number of channels
        if is_input_device:
            num_channels = input_channels
            print(f"\nUsing {num_channels} channels for input streaming")
        else:
            num_channels = output_channels
            print(f"\nUsing {num_channels} channels for output monitoring")
        
        # Cap at 2 channels for simplicity
        num_channels = min(num_channels, 2)
        
        debug_log(f"Creating streamer with device_id={device_id}, name={device_name}, is_input={is_input_device}")
        
        # Create and start the streamer with the shared timer
        streamer = LSLAudioStreamer(
            device_id=device_id,
            device_name=device_name,
            channel_name=stream_name,
            silence_threshold=0.01,
            shared_timer=shared_timer,
            num_channels=num_channels,
            is_input_device=is_input_device  # Pass whether this is an input device
        )
        
        # Start the streamer and let it run until the stop event is set
        try:
            debug_log(f"Starting streamer for {stream_name}")
            streamer.start()
        except Exception as e:
            print(f"Error in streamer thread: {e}")
            debug_log(f"Error in streamer thread: {e}")
            debug_log(f"Exception: {traceback.format_exc()}")
        finally:
            print(f"Streamer thread for {stream_name} exiting")
            debug_log(f"Streamer thread for {stream_name} exiting")
    
    except Exception as e:
        print(f"Critical error setting up audio streamer: {e}")
        debug_log(f"Critical error setting up audio streamer: {e}")
        debug_log(f"Exception: {traceback.format_exc()}")
    finally:
        p.terminate()

def main():
    """Main function with enhanced diagnostics."""
    debug_log("Starting main function")
    print("Starting audio streaming with enhanced diagnostics...")
    print(f"Debug mode: {'ON' if DEBUG else 'OFF'}")
    
    try:
        # List available audio devices
        debug_log("Calling list_audio_devices()")
        list_audio_devices()
        
        # Find Komplete output devices
        debug_log("Calling find_komplete_output_devices()")
        output1_2, output3_4 = find_komplete_output_devices()
        
        # Make sure we found the devices
        if not output1_2:
            print("Could not find Komplete Output 1/2 device! Check device names and connections.")
            debug_log("Komplete Output 1/2 device not found!")
            return
            
        if not output3_4:
            print("Could not find Komplete Output 3/4 device! Check device names and connections.")
            debug_log("Komplete Output 3/4 device not found!")
            return
            
        print(f"\nFound Output 1/2: ID {output1_2[0]}, {output1_2[1]}, Channels: {output1_2[2]}")
        print(f"Found Output 3/4: ID {output3_4[0]}, {output3_4[1]}, Channels: {output3_4[2]}")
        
        # Create a shared timer to synchronize both streamers
        debug_log("Creating shared timer and stop event")
        shared_timer = {
            'start_time': None,
            'running': False,
            'lock': threading.Lock()
        }
        
        # Create a shared stop event
        stop_event = threading.Event()
        
        # Create threads for each output device
        threads = []
        
        print("\nStarting audio streamers for both output channels...")
        print("Both streams will start simultaneously when audio is detected on either channel.")
        print("Both streams will run for exactly 2 minutes from first audio detection.")
        
        # Start streamer for Output 1/2
        debug_log("Creating thread for Output 1/2")
        t1 = threading.Thread(
            target=run_output_streamer,
            args=(output1_2, shared_timer, stop_event),
            name="Output_1_2_Thread"
        )
        t1.daemon = True
        debug_log("Starting thread for Output 1/2")
        t1.start()
        threads.append(t1)
        debug_log("Thread for Output 1/2 started")
        
        # Start streamer for Output 3/4
        debug_log("Creating thread for Output 3/4")
        t2 = threading.Thread(
            target=run_output_streamer,
            args=(output3_4, shared_timer, stop_event),
            name="Output_3_4_Thread"
        )
        t2.daemon = True
        debug_log("Starting thread for Output 3/4")
        t2.start()
        threads.append(t2)
        debug_log("Thread for Output 3/4 started")
        
        print("\nBoth streamers started and synchronized. Press Ctrl+C to stop all streamers.")
        debug_log("Both streamer threads are now running")
        
        # Wait for all threads to finish or until interrupted
        try:
            debug_log("Main thread waiting for worker threads")
            for i, t in enumerate(threads):
                debug_log(f"Joining thread {i+1}: {t.name}")
                t.join()
                debug_log(f"Thread {i+1} joined")
        except KeyboardInterrupt:
            print("\nReceived keyboard interrupt. Stopping all streamers...")
            debug_log("Received keyboard interrupt, stopping all threads")
            stop_event.set()
            
            # Give threads a moment to clean up
            debug_log("Waiting for threads to clean up")
            time.sleep(1)
            
    except Exception as e:
        print(f"\nError running script: {e}")
        debug_log(f"Error in main function: {e}")
        debug_log(f"Exception: {traceback.format_exc()}")
        traceback.print_exc()
        
    print("\nScript execution complete.")
    debug_log("Main function completed")

# For Jupyter notebook compatibility
if __name__ == "__main__" or 'get_ipython' in globals():
    debug_log("Script starting as main or in Jupyter")
    # Check if running in a Jupyter notebook
    try:
        debug_log("Checking for Jupyter environment")
        get_ipython  # This will raise NameError if not in Jupyter
        debug_log("Running in Jupyter environment")
        print("Running in Jupyter notebook environment")
        main()
    except NameError:
        # Not in a Jupyter notebook, run normally
        debug_log("Not running in Jupyter, executing main normally")
        main()
else:
    debug_log("Script imported as a module, not executing main()")
```
