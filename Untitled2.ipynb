```python
import sounddevice as sd
import soundfile as sf
import numpy as np
import time

# Path to the looming stimulus file
LOOMING_FILE = "C:\\Users\\cogpsy-vrlab\\Documents\\GitHub\\BreathingSpace\\Level1_AudioGeneration\\ExperimentAudio\\participant_0_design_looming.wav"

# Woojer device ID (change this if needed)
WOOJER_ID = 7  # ID 7: Speakers (Woojer Strap Edge)

def play_looming():
    """Play looming stimulus through Woojer"""
    # Load the audio file
    print(f"Loading audio: {LOOMING_FILE}")
    audio_data, sample_rate = sf.read(LOOMING_FILE)
    
    # Convert to mono if stereo
    if len(audio_data.shape) > 1:
        audio_data = np.mean(audio_data, axis=1)
    
    # Reshape for output and normalize volume
    audio_data = audio_data.reshape(-1, 1)
    audio_data = audio_data / np.max(np.abs(audio_data)) * 0.9
    
    # Play the audio
    print(f"Playing through Woojer (device ID: {WOOJER_ID})")
    sd.play(audio_data, sample_rate, device=WOOJER_ID)
    sd.wait()
    
    print("Playback complete")

if __name__ == "__main__":
    play_looming()
```
