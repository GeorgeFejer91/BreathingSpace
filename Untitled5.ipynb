```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-
"""
Compatible PPS Design and Audio Generator

This script avoids using pandas to prevent NumPy 2.x compatibility issues.
It generates 2-channel stereo audio files for PPS experiments where:
- Left channel: Looming stimuli + box breathing
- Right channel: Tactile stimuli

All outputs saved to specified folder.
"""

import os
import numpy as np
import csv
import re
import datetime
import traceback
import random
import time
from collections import defaultdict, Counter

# ----------------------------------------------------------------------
# CONFIGURATION PARAMETERS
# ----------------------------------------------------------------------
CONFIG = {
    # Paths
    'output_dir': r"C:\Users\cogpsy-vrlab\Desktop\audiogeneration",
    'stimulus_dir': r"C:\Users\cogpsy-vrlab\Documents\GitHub\BreathingSpace\Level0_StimulusFiles",
    
    # Audio settings
    'sample_rate': 48000,  # Fixed sample rate for Komplete Audio compatibility
    
    # Experiment design parameters
    'repetitions': 5,       # Trials per SOA per condition
    'practice_trials': 3,   # Number of practice trials at beginning
    'num_participants': 5,  # Number of participant designs to generate
    
    # Stimulus parameters
    'soa_conditions_ms': [50, 100, 200, 300],  # SOA values in milliseconds
    'jitter_range_ms': [-20, 20],              # Random jitter range in ms
    'stimulus_types': ['white', 'pink', 'blue', 'brown'],  # Types of looming stimuli
    
    # Updated stimulus file paths
    'looming_dir': r"C:\Users\cogpsy-vrlab\Documents\GitHub\BreathingSpace\Level0_StimulusFiles\LoomingStimuli",  # Subdirectory with looming files
    'looming_file_pattern': 'frontal_looming_{}_noise.wav',  # File naming pattern
    
    # Debug mode
    'debug_mode': True      # Enable detailed output
}

# Ensure output directory exists
os.makedirs(CONFIG['output_dir'], exist_ok=True)

# ----------------------------------------------------------------------
# DESIGN GENERATOR CLASS
# ----------------------------------------------------------------------
class PPSDesignGenerator:
    """Generates counterbalanced participant-specific experimental designs."""
    
    def __init__(self, config=None):
        self.config = config or CONFIG
        self.output_dir = self.config['output_dir']
        
        # Define trial types
        self.trial_types = ['baseline', 'catch', 'inhale', 'exhale']
        
        # Track total trials
        self.total_trials_per_participant = self._calculate_total_trials()
        
        # Track last trial properties to avoid consecutive identical trials
        self.last_trial_type = None
        self.last_stimulus_type = None
        
        # Print initialization info if debug mode
        if self.config['debug_mode']:
            print(f"Initialized PPS Design Generator")
            print(f"Output directory: {self.output_dir}")
            print(f"Total trials per participant: {self.total_trials_per_participant}")
    
    def _calculate_total_trials(self):
        """Calculate total number of trials needed based on config"""
        # Calculate experimental trials
        # 4 trial types x len(SOAs) x repetitions
        trial_types_count = len(self.trial_types)
        soa_count = len(self.config['soa_conditions_ms'])
        repetitions = self.config['repetitions']
        
        experimental_trials = trial_types_count * soa_count * repetitions
        practice_trials = self.config['practice_trials']
        
        return experimental_trials + practice_trials
    
    def generate_counterbalanced_design(self, participant_id):
        """Generate counterbalanced design for a participant"""
        # Create practice trials
        practice_trials = self._create_practice_trials()
        
        # Create experimental trials
        experimental_trials = self._create_experimental_trials(participant_id)
        
        # Combine and create final list
        all_trials = practice_trials + experimental_trials
        
        # Assign trial numbers
        for i, trial in enumerate(all_trials):
            trial['trial_number'] = i + 1
        
        return all_trials
    
    def _create_practice_trials(self):
        """Create practice trials with balanced conditions"""
        practice_trials = []
        practice_count = self.config['practice_trials']
        
        # Sample from real conditions for practice
        trial_types = random.sample(self.trial_types, min(practice_count, len(self.trial_types)))
        
        # Fill remaining with random selections if needed
        while len(trial_types) < practice_count:
            trial_types.append(random.choice(self.trial_types))
        
        # Create practice trials
        for i in range(practice_count):
            trial_type = trial_types[i]
            soa = random.choice(self.config['soa_conditions_ms'])
            stimulus_type = random.choice(self.config['stimulus_types'])
            
            trial = {
                'trial_type': 'practice',
                'condition': trial_type,
                'soa_ms': soa,
                'stimulus_type': stimulus_type,
                'jitter_ms': self._get_random_jitter(),
                'breathphase': random.choice(['inhale', 'exhale', 'hold']),
            }
            practice_trials.append(trial)
        
        return practice_trials
    
    def _create_experimental_trials(self, participant_id):
        """Create counterbalanced experimental trials"""
        experimental_trials = []
        
        # Create a balanced set of all conditions
        conditions = []
        for trial_type in self.trial_types:
            for soa in self.config['soa_conditions_ms']:
                for _ in range(self.config['repetitions']):
                    conditions.append((trial_type, soa))
        
        # Shuffle conditions for this participant (seeded by participant_id)
        random.seed(1000 + participant_id)
        random.shuffle(conditions)
        
        # Reset tracking variables
        self.last_trial_type = None
        self.last_stimulus_type = None
        
        # Assign conditions to trials with constraints
        shuffled_trials = []
        
        # Convert conditions to list of trials first
        for trial_type, soa in conditions:
            # Pick a stimulus type avoiding repetition
            stimulus_type = self._get_diverse_stimulus_type()
            
            trial = {
                'trial_type': trial_type,
                'soa_ms': soa,
                'stimulus_type': stimulus_type,
                'jitter_ms': self._get_random_jitter(),
                'breathphase': None,  # Will be assigned later
            }
            shuffled_trials.append(trial)
            
            # Update tracking variables
            self.last_trial_type = trial_type
            self.last_stimulus_type = stimulus_type
        
        # Apply constraints to prevent sequences of the same trial type
        experimental_trials = self._apply_sequencing_constraints(shuffled_trials)
        
        # Assign breath phases ensuring balance
        experimental_trials = self._assign_breath_phases(experimental_trials, participant_id)
        
        return experimental_trials
    
    def _get_diverse_stimulus_type(self):
        """Get a stimulus type different from the last one"""
        available_types = self.config['stimulus_types']
        
        if self.last_stimulus_type is None:
            return random.choice(available_types)
        
        # Avoid same stimulus type twice in a row
        remaining_types = [t for t in available_types if t != self.last_stimulus_type]
        return random.choice(remaining_types)
    
    def _get_random_jitter(self):
        """Get a random jitter value within the configured range"""
        min_jitter, max_jitter = self.config['jitter_range_ms']
        return random.randint(min_jitter, max_jitter)
    
    def _apply_sequencing_constraints(self, trials):
        """Apply constraints to prevent sequences of the same trial type"""
        max_attempts = 1000
        count = 0
        
        while count < max_attempts:
            # Check for consecutive trials of same type
            has_consecutive = False
            for i in range(1, len(trials)):
                if trials[i]['trial_type'] == trials[i-1]['trial_type']:
                    has_consecutive = True
                    break
            
            if not has_consecutive:
                break
            
            # Shuffle and try again
            random.shuffle(trials)
            count += 1
        
        return trials
    
    def _assign_breath_phases(self, trials, participant_id):
        """Assign breath phases to trials with balance constraints"""
        # Set a different seed for breath phase assignment
        random.seed(2000 + participant_id)
        
        # Group trials by type
        trial_groups = defaultdict(list)
        for trial in trials:
            trial_groups[trial['trial_type']].append(trial)
        
        # Deterministic assignment for inhale/exhale trial types
        for trial in trial_groups.get('inhale', []):
            trial['breathphase'] = 'inhale'
        
        for trial in trial_groups.get('exhale', []):
            trial['breathphase'] = 'exhale'
        
        # Balanced random assignment for baseline/catch trials
        flexible_trials = []
        for trial_type in ['baseline', 'catch']:
            flexible_trials.extend(trial_groups.get(trial_type, []))
        
        # Determine counts for balanced assignment
        target_count = len(flexible_trials) // 2
        
        # Shuffle flexible trials
        random.shuffle(flexible_trials)
        
        # Assign first half to inhale, second to exhale
        for i, trial in enumerate(flexible_trials):
            trial['breathphase'] = 'inhale' if i < target_count else 'exhale'
        
        # Return all trials
        return trials
    
    def finalize_design_csv(self, trials):
        """Add timestamps and finalize the design"""
        # Define base trial timing parameters
        base_time_sec = 0  # Start time
        trial_spacing_sec = 15  # Seconds between trial onsets
        
        # Add timestamps to each trial
        for i, trial in enumerate(trials):
            # Calculate base looming stimulus time
            looming_time_sec = base_time_sec + (i * trial_spacing_sec)
            
            # Format as MM:SS.S
            looming_minutes = int(looming_time_sec // 60)
            looming_seconds = looming_time_sec % 60
            looming_timestamp = f"{looming_minutes}:{looming_seconds:.1f}"
            
            # Calculate tactile stimulus time based on SOA
            soa_sec = trial['soa_ms'] / 1000.0
            jitter_sec = trial['jitter_ms'] / 1000.0
            
            tactile_time_sec = looming_time_sec + soa_sec + jitter_sec
            
            # Format as MM:SS.S
            tactile_minutes = int(tactile_time_sec // 60)
            tactile_seconds = tactile_time_sec % 60
            tactile_timestamp = f"{tactile_minutes}:{tactile_seconds:.1f}"
            
            # Add timestamps to trial
            trial['looming_stimulus_timestamp'] = looming_timestamp
            trial['tactile_stimulus_timestamp'] = tactile_timestamp
        
        return trials
    
    def analyze_design(self, trials):
        """Analyze the design for balance and randomization quality (no pandas)"""
        if not self.config['debug_mode']:
            return
        
        print(f"\nDesign analysis:")
        
        # Separate practice and experimental trials
        exp_trials = [t for t in trials if t['trial_type'] != 'practice']
        experimental_count = len(exp_trials)
        
        # 1. Count of each trial type
        print(f"Trial type distribution (excluding practice):")
        type_counter = Counter(t['trial_type'] for t in exp_trials)
        for trial_type, count in sorted(type_counter.items()):
            percent = (count / experimental_count) * 100
            print(f"  {trial_type}: {count} trials ({percent:.1f}%)")
        
        # 2. Count of each SOA
        print(f"\nSOA distribution (excluding practice):")
        soa_counter = Counter(t['soa_ms'] for t in exp_trials)
        for soa, count in sorted(soa_counter.items()):
            percent = (count / experimental_count) * 100
            print(f"  {soa} ms: {count} trials ({percent:.1f}%)")
        
        # 3. Count of each stimulus type
        print(f"\nStimulus type distribution (excluding practice):")
        stim_counter = Counter(t['stimulus_type'] for t in exp_trials)
        for stim, count in sorted(stim_counter.items()):
            percent = (count / experimental_count) * 100
            print(f"  {stim}: {count} trials ({percent:.1f}%)")
        
        # 4. SOA by trial type
        print(f"\nSOA by trial type (excluding practice):")
        for trial_type in sorted(type_counter.keys()):
            subset = [t for t in exp_trials if t['trial_type'] == trial_type]
            soa_counter = Counter(t['soa_ms'] for t in subset)
            print(f"  {trial_type.upper()}:")
            for soa, count in sorted(soa_counter.items()):
                percent = (count / len(subset)) * 100
                print(f"    {soa} ms: {count} trials ({percent:.1f}%)")
        
        # 5. Breath phase distribution
        print("\nBreath phase distribution (excluding practice):")
        phase_counter = Counter(t['breathphase'] for t in exp_trials)
        for phase, count in sorted(phase_counter.items()):
            percent = (count / experimental_count) * 100
            print(f"  {phase}: {count} trials ({percent:.1f}%)")
        
        # 6. Cross-tabulation of trial types by breath phase
        print("\nTrial types by breath phase (COUNTERBALANCE CHECK):")
        
        # Simple cross-tab calculation
        cross_tab = {}
        for trial_type in sorted(type_counter.keys()):
            cross_tab[trial_type] = {}
            for phase in sorted(phase_counter.keys()):
                cross_tab[trial_type][phase] = sum(1 for t in exp_trials 
                                                 if t['trial_type'] == trial_type and t['breathphase'] == phase)
        
        # Print as table
        phases = sorted(phase_counter.keys())
        header = "Trial Type | " + " | ".join(phases) + " |"
        print(header)
        print("-" * len(header))
        
        for trial_type in sorted(type_counter.keys()):
            row = f"{trial_type:10} |"
            for phase in phases:
                row += f" {cross_tab[trial_type][phase]:3d} |"
            print(row)
    
    def save_design_to_csv(self, trials, participant_id):
        """Save design to CSV file without pandas"""
        out_path = os.path.join(self.output_dir, f"participant_{participant_id}_design.csv")
        
        # Determine fieldnames from the first trial
        if trials:
            fieldnames = sorted(trials[0].keys())
        else:
            fieldnames = []
        
        # Write to CSV
        with open(out_path, 'w', newline='') as csvfile:
            writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
            writer.writeheader()
            writer.writerows(trials)
        
        if self.config['debug_mode']:
            print(f"Saved participant {participant_id} design to {out_path}")
        
        return out_path
    
    def process_participant(self, participant_id):
        """Process a single participant: generate design and save CSV."""
        try:
            if self.config['debug_mode']:
                print(f"\n=== Processing participant {participant_id} ===")
            
            # Generate design
            trials = self.generate_counterbalanced_design(participant_id)
            
            # Finalize design with timestamps
            trials = self.finalize_design_csv(trials)
            
            # Analyze design if in debug mode
            self.analyze_design(trials)
            
            # Save CSV
            out_path = self.save_design_to_csv(trials, participant_id)
            
            return True, out_path, trials
            
        except Exception as e:
            print(f"Error processing participant {participant_id}: {e}")
            traceback.print_exc()
            return False, None, None
    
    def run(self, num_participants=None):
        """Generate designs for all participants"""
        if num_participants is None:
            num_participants = self.config['num_participants']
        
        participant_ids = list(range(num_participants))
        
        print(f"Generating designs for {num_participants} participants...")
        
        results = {}
        designs = {}
        
        for pid in participant_ids:
            success, path, trials = self.process_participant(pid)
            results[pid] = {'success': success, 'csv_path': path}
            designs[pid] = trials
        
        return results, designs

# ----------------------------------------------------------------------
# AUDIO GENERATOR CLASS
# ----------------------------------------------------------------------
class PPSAudioGenerator:
    """Generates participant-specific 2-channel stereo audio files."""
    
    def __init__(self, config=None, designs=None):
        self.config = config or CONFIG
        self.output_dir = self.config['output_dir']
        self.stimulus_dir = self.config['stimulus_dir']
        self.sample_rate = self.config['sample_rate']
        self.designs = designs or {}
        
        # Load sound file module outside of NumPy dependency
        import soundfile as sf
        self.sf = sf
        
        # Create base audio (box breathing rhythm)
        self.base_audio = self._create_base_audio()
        
        # Load stimulus files
        self.looming_stimuli = self._load_looming_stimuli()
        self.tactile_stimulus = self._load_tactile_stimulus()
    
    def _create_base_audio(self):
        """Create a simple box breathing rhythm audio as base for looming channel"""
        # Simple 4-4-4-4 box breathing rhythm
        # Just use silence for simplicity, duration based on total trial count
        try:
            import numpy as np
            duration_sec = 15 * self.config['num_participants'] * self._estimate_trial_count() 
            samples = int(duration_sec * self.sample_rate)
            return np.zeros(samples, dtype=np.float32)
        except ImportError:
            # If NumPy fails, create a list of zeros
            duration_sec = 15 * self.config['num_participants'] * self._estimate_trial_count()
            samples = int(duration_sec * self.sample_rate)
            return [0.0] * samples
    
    def _estimate_trial_count(self):
        """Estimate trial count based on config"""
        # For practice trials
        practice_count = self.config['practice_trials']
        
        # For experimental trials (4 trial types * SOAs * repetitions)
        experimental_count = 4 * len(self.config['soa_conditions_ms']) * self.config['repetitions']
        
        return practice_count + experimental_count
    
    def _load_looming_stimuli(self):
        """Load all looming stimulus files."""
        import numpy as np
        looming_stimuli = {}
        
        # Get looming directory and file pattern from config
        looming_dir = self.config.get('looming_dir', self.stimulus_dir)
        file_pattern = self.config.get('looming_file_pattern', 'frontal_looming_{}_noise.wav')
        
        stim_config = self.config['stimulus_types']
        for stim_type in stim_config:
            # Construct filenames based on type
            filename = file_pattern.format(stim_type)
            file_path = os.path.join(looming_dir, filename)
            
            print(f"Looking for looming stimulus: {file_path}")
            
            if os.path.exists(file_path):
                try:
                    data, sr = self.sf.read(file_path)
                    
                    # Convert to mono if stereo
                    if len(data.shape) > 1 and data.shape[1] > 1:
                        data = data[:, 0]  # Take left channel
                    
                    # Resample if needed
                    if sr != self.sample_rate:
                        # Simple resampling
                        try:
                            ratio = self.sample_rate / sr
                            new_length = int(len(data) * ratio)
                            old_time = np.arange(len(data)) / sr
                            new_time = np.arange(new_length) / self.sample_rate
                            data = np.interp(new_time, old_time, data)
                        except Exception as e:
                            print(f"Error resampling: {e}")
                            # Fallback resampling with simple interpolation
                            ratio = self.sample_rate / sr
                            new_length = int(len(data) * ratio)
                            data = self._simple_resample(data, sr, self.sample_rate)
                    
                    looming_stimuli[stim_type] = {
                        'data': data, 
                        'sample_rate': self.sample_rate,
                        'duration': len(data) / self.sample_rate,
                    }
                    print(f"Successfully loaded looming stimulus: {stim_type}")
                except Exception as e:
                    print(f"Error loading looming stimulus {stim_type}: {e}")
                    traceback.print_exc()
            else:
                print(f"Looming stimulus file not found: {file_path}")
        
        return looming_stimuli
    
    def _simple_resample(self, data, orig_sr, target_sr):
        """Very simple resampling without NumPy dependencies"""
        # Calculate ratio
        ratio = target_sr / orig_sr
        new_length = int(len(data) * ratio)
        
        # Create new array
        resampled = [0.0] * new_length
        
        # Simple linear interpolation
        for i in range(new_length):
            src_idx = i / ratio
            src_idx_floor = int(src_idx)
            src_idx_ceil = min(len(data) - 1, src_idx_floor + 1)
            fraction = src_idx - src_idx_floor
            
            # Linear interpolation
            resampled[i] = data[src_idx_floor] * (1 - fraction) + data[src_idx_ceil] * fraction
        
        return resampled
    
    def _load_tactile_stimulus(self):
        """Load tactile stimulus file."""
        import numpy as np
        
        # Look for tactile stimulus in multiple locations
        possible_paths = [
            os.path.join(self.stimulus_dir, 'tactile_stimulus.wav'),
            os.path.join(self.stimulus_dir, 'TactileStimuli', 'tactile_stimulus.wav'),
            os.path.join(self.config.get('looming_dir', self.stimulus_dir), 'tactile_stimulus.wav')
        ]
        
        tactile_path = None
        for path in possible_paths:
            if os.path.exists(path):
                tactile_path = path
                break
        
        if tactile_path and os.path.exists(tactile_path):
            print(f"Found tactile stimulus: {tactile_path}")
            try:
                data, sr = self.sf.read(tactile_path)
                
                # Convert to mono if stereo
                if len(data.shape) > 1 and data.shape[1] > 1:
                    data = data[:, 0]  # Take left channel
                
                # Resample if needed
                if sr != self.sample_rate:
                    try:
                        # Simple resampling
                        ratio = self.sample_rate / sr
                        new_length = int(len(data) * ratio)
                        old_time = np.arange(len(data)) / sr
                        new_time = np.arange(new_length) / self.sample_rate
                        data = np.interp(new_time, old_time, data)
                    except Exception as e:
                        print(f"Error resampling: {e}")
                        # Fallback resampling
                        data = self._simple_resample(data, sr, self.sample_rate)
                
                tactile_info = {
                    'data': data,
                    'sample_rate': self.sample_rate,
                    'duration': len(data) / self.sample_rate,
                }
                print(f"Successfully loaded tactile stimulus")
                return tactile_info
            except Exception as e:
                print(f"Error loading tactile stimulus: {e}")
                traceback.print_exc()
        else:
            print(f"Tactile stimulus file not found in any expected location. Creating default beep.")
        
        # Return a simple default tactile stimulus if file not loaded (short beep)
        try:
            import numpy as np
            duration = 0.1  # 100 ms
            samples = int(duration * self.sample_rate)
            t = np.linspace(0, duration, samples, endpoint=False)
            beep = 0.5 * np.sin(2 * np.pi * 440 * t)  # 440 Hz sine wave
            
            # Add simple envelope
            envelope = np.ones_like(beep)
            fade_samples = int(0.01 * self.sample_rate)  # 10ms fade
            envelope[:fade_samples] = np.linspace(0, 1, fade_samples)
            envelope[-fade_samples:] = np.linspace(1, 0, fade_samples)
            beep = beep * envelope
        except ImportError:
            # Fallback without NumPy
            duration = 0.1  # 100 ms
            samples = int(duration * self.sample_rate)
            beep = [0.0] * samples
            # Create a simple beep without NumPy
            for i in range(samples):
                t = i / self.sample_rate
                beep[i] = 0.5 * math.sin(2 * math.pi * 440 * t)
                
                # Apply simple envelope
                if i < samples * 0.1:  # First 10%
                    beep[i] *= i / (samples * 0.1)
                elif i > samples * 0.9:  # Last 10%
                    beep[i] *= (samples - i) / (samples * 0.1)
        
        return {
            'data': beep,
            'sample_rate': self.sample_rate,
            'duration': duration,
        }
    
    def parse_timestamp(self, timestamp):
        """Parse timestamp in format MM:SS.S to milliseconds."""
        if not timestamp:
            return None
            
        match = re.match(r'(\d+):(\d+\.\d+)', timestamp)
        if match:
            minutes, seconds = match.groups()
            return (int(minutes) * 60 + float(seconds)) * 1000
        return None
    
    def create_participant_audio(self, participant_id, trials):
        """
        Create participant-specific 2-channel stereo audio file with:
        - Left channel: Looming stimuli overlaid on breathing guide
        - Right channel: Tactile stimuli
        """
        import numpy as np
        print(f"Creating 2-channel audio for participant {participant_id}")
        
        # Separate looming and tactile trials
        looming_trials = []
        tactile_trials = []
        
        for trial in trials:
            trial_num = trial['trial_number']
            trial_type = trial['trial_type']
            
            # Skip practice trials
            if trial_type == 'practice':
                continue
            
            # Process looming stimulus for non-baseline trials
            if trial_type != 'baseline' and 'looming_stimulus_timestamp' in trial:
                looming_ts = trial['looming_stimulus_timestamp']
                looming_ms = self.parse_timestamp(looming_ts)
                
                if looming_ms is not None:
                    looming_trial = {
                        'trial_number': trial_num,
                        'trial_type': trial_type,
                        'timestamp_ms': looming_ms,
                        'stimulus_type': trial.get('stimulus_type', 'white'),
                    }
                    looming_trials.append(looming_trial)
            
            # Process tactile stimulus for non-catch trials
            if trial_type != 'catch' and 'tactile_stimulus_timestamp' in trial:
                tactile_ts = trial['tactile_stimulus_timestamp']
                tactile_ms = self.parse_timestamp(tactile_ts)
                
                if tactile_ms is not None:
                    tactile_trial = {
                        'trial_number': trial_num,
                        'trial_type': trial_type,
                        'timestamp_ms': tactile_ms,
                    }
                    tactile_trials.append(tactile_trial)
        
        # Sort by timestamp
        looming_trials.sort(key=lambda x: x['timestamp_ms'])
        tactile_trials.sort(key=lambda x: x['timestamp_ms'])
        
        # Get length of base audio or calculate based on last timestamp
        if self.base_audio is not None:
            looming_audio = np.copy(self.base_audio)
            duration_samples = len(looming_audio)
        else:
            # Calculate based on last timestamp plus margin
            last_looming = max(t['timestamp_ms'] for t in looming_trials) if looming_trials else 0
            last_tactile = max(t['timestamp_ms'] for t in tactile_trials) if tactile_trials else 0
            last_timestamp = max(last_looming, last_tactile)
            
            # Add 30 seconds margin
            duration_sec = (last_timestamp / 1000) + 30
            duration_samples = int(self.sample_rate * duration_sec)
            
            # Create silent base audio
            looming_audio = np.zeros(duration_samples, dtype=np.float32)
        
        # Create empty array for tactile audio (right channel)
        tactile_audio = np.zeros_like(looming_audio)
        
        # Insert looming stimuli into left channel
        looming_inserted = 0
        for trial in looming_trials:
            trial_num = trial['trial_number']
            timestamp_ms = trial['timestamp_ms']
            stim_type = trial['stimulus_type']
            
            # Convert timestamp to sample index
            start_sample = int(timestamp_ms * self.sample_rate / 1000)
            
            # Get stimulus data
            if stim_type in self.looming_stimuli:
                stim_data = self.looming_stimuli[stim_type]['data']
                end_sample = start_sample + len(stim_data)
                
                # Check if it fits within array bounds
                if end_sample <= len(looming_audio):
                    # Insert stimulus
                    looming_audio[start_sample:end_sample] += stim_data
                    looming_inserted += 1
                else:
                    print(f"Warning: Trial {trial_num}: Looming stimulus exceeds audio length")
            else:
                print(f"Warning: Trial {trial_num}: Looming stimulus type '{stim_type}' not found")
        
        # Insert tactile stimuli into right channel
        tactile_inserted = 0
        for trial in tactile_trials:
            trial_num = trial['trial_number']
            timestamp_ms = trial['timestamp_ms']
            
            # Convert timestamp to sample index
            start_sample = int(timestamp_ms * self.sample_rate / 1000)
            
            # Get stimulus data
            if self.tactile_stimulus:
                stim_data = self.tactile_stimulus['data']
                end_sample = start_sample + len(stim_data)
                
                # Check if it fits within array bounds
                if end_sample <= len(tactile_audio):
                    # Insert stimulus
                    tactile_audio[start_sample:end_sample] += stim_data
                    tactile_inserted += 1
                else:
                    print(f"Warning: Trial {trial_num}: Tactile stimulus exceeds audio length")
        
        # Normalize audio if needed (prevent clipping)
        max_looming = np.max(np.abs(looming_audio))
        if max_looming > 0.95:
            scale_factor = 0.95 / max_looming
            looming_audio *= scale_factor
        
        max_tactile = np.max(np.abs(tactile_audio))
        if max_tactile > 0.95:
            scale_factor = 0.95 / max_tactile
            tactile_audio *= scale_factor
        
        # Combine into 2-channel stereo audio
        stereo_audio = np.column_stack((looming_audio, tactile_audio)).astype(np.float32)
        
        # Save 2-channel audio file
        filename = f"participant_{participant_id}_design_stereo.wav"
        output_path = os.path.join(self.output_dir, filename)
        
        try:
            # Write stereo WAV file
            self.sf.write(output_path, stereo_audio, self.sample_rate)
            print(f"Created audio file: {filename} (looming: {looming_inserted}, tactile: {tactile_inserted})")
            return output_path
            
        except Exception as e:
            print(f"Error saving audio file: {e}")
            traceback.print_exc()
            return None
    
    def process_participants(self):
        """Process all participants using design dataframes from self.designs"""
        results = {}
        
        for participant_id, trials in self.designs.items():
            print(f"Processing audio for participant {participant_id}")
            output_path = self.create_participant_audio(participant_id, trials)
            
            results[participant_id] = {
                'success': output_path is not None,
                'audio_path': output_path
            }
        
        return results

# ----------------------------------------------------------------------
# MAIN FUNCTION - Can be run in Jupyter or as standalone script
# ----------------------------------------------------------------------
def run_combined_generator(config=None):
    """Run the combined design and audio generator"""
    if config is None:
        config = CONFIG
    
    print(f"\n===== STARTING COMBINED PPS DESIGN AND AUDIO GENERATOR =====")
    print(f"Output directory: {config['output_dir']}")
    print(f"Sample rate: {config['sample_rate']} Hz")
    print(f"Participants: {config['num_participants']}")
    
    # Check if looming stimuli directory exists
    looming_dir = config.get('looming_dir', config['stimulus_dir'])
    if not os.path.exists(looming_dir):
        print(f"WARNING: Looming stimuli directory not found: {looming_dir}")
        print(f"Please make sure the directory exists and contains the appropriate .wav files.")
    
    # Step 1: Generate participant designs
    print(f"\n----- GENERATING PARTICIPANT DESIGNS -----")
    design_generator = PPSDesignGenerator(config)
    design_results, designs = design_generator.run()
    
    # Step 2: Generate audio files
    print(f"\n----- GENERATING AUDIO FILES -----")
    audio_generator = PPSAudioGenerator(config, designs)
    audio_results = audio_generator.process_participants()
    
    # Print summary
    print(f"\n===== GENERATION COMPLETE =====")
    
    # Count successes
    design_success = sum(1 for r in design_results.values() if r['success'])
    audio_success = sum(1 for r in audio_results.values() if r['success'])
    
    print(f"Designs created: {design_success}/{len(design_results)}")
    print(f"Audio files created: {audio_success}/{len(audio_results)}")
    print(f"All files saved to: {config['output_dir']}")
    
    # Print instructions for playback
    print("\nPlayback Instructions:")
    print("1. The stereo WAV files contain:")
    print("   - Left channel: Looming stimuli + box breathing")
    print("   - Right channel: Tactile stimuli")
    print("2. Use the playback script to route channels appropriately")
    
    return {
        'design_results': design_results,
        'audio_results': audio_results
    }

# Run if executed as script
if __name__ == "__main__":
    run_combined_generator()
```
