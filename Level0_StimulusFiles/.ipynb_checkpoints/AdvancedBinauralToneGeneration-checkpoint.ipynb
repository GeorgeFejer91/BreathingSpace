```python
class BiAuralConfig:
    """Central configuration class for binaural rendering settings"""
    
    class MovementType(Enum):
        LINEAR = "linear"
        CIRCULAR = "circular" 
        SPIRAL = "spiral"
        RANDOM_WALK = "random_walk"
        STATIONARY_STOPS = "stationary_stops"
    
    class HRTFInterpolation(Enum):
        NEAREST = "nearest"
        BILINEAR = "bilinear"
    
    class RoomType(Enum):
        ANECHOIC = "anechoic"
        SHOE_BOX = "shoe_box"
        SPHERICAL = "spherical"  # Added spherical room type
        
    def __init__(self):
        # Source parameters
        self.movement_type = self.MovementType.CIRCULAR
        self.duration = 8.0        # Duration in seconds
        self.start_position = [-1.0, 1.0, 0.0]  # Starting position [x, y, z] in meters
        self.end_position = [1.0, 1.0, 0.0]     # Ending position for linear movement
        self.orbit_radius = 1.0    # Radius for circular movement
        self.orbit_height = 0.0    # Height for circular movement
        self.min_distance = 0.2    # Minimum distance from head in meters
        self.orbit_speed = 0.5     # Orbits per second
        
        # Head and listener parameters
        self.head_radius = 0.0875  # Average human head radius in meters
        self.head_position = [0.0, 0.0, 0.0]  # Head position [x, y, z] in meters
        self.head_orientation = [0.0, 0.0, 0.0]  # Head orientation [yaw, pitch, roll] in degrees
        self.ear_spacing = 0.165   # Distance between ears in meters
        
        # Audio parameters
        self.fs = 48000             # Sampling frequency in Hz
        self.speed_of_sound = 343.0 # Speed of sound in m/s
        self.source_type = "pink"   # Source signal type: "pink", "white", "speech", "impulse"
        
        # HRTF parameters
        self.hrtf_interpolation = self.HRTFInterpolation.BILINEAR  # HRTF interpolation method
        self.enable_near_field = True  # Apply near-field HRTF compensation
        
        # Room parameters
        self.room_type = self.RoomType.SPHERICAL  # Default to spherical room
        self.room_radius = 2.0      # Radius of the spherical room in meters
        self.wall_absorption = 0.3  # Absorption coefficient for the spherical room
        self.max_order = 4          # Maximum reflection order
        self.ray_tracing = True     # Use ray tracing for higher-order reflections
        self.num_rays = 10000       # Number of rays for ray tracing
        self.scattering = 0.1       # Scattering coefficient
        self.air_absorption = True  # Model air absorption
        
        # Rendering parameters
        self.enable_doppler = True  # Simulate doppler effect
        self.enable_distance_attenuation = True  # Apply 1/r distance attenuation
        self.direct_gain = 1.0      # Gain factor for direct sound
        self.reverb_gain = 0.6      # Gain factor for reverberant sound
        self.block_size = 1024      # Block size for processing
        
        # Output parameters
        self.apply_loudness_normalization = True
        self.headphone_compensation = True
        self.output_dir = os.path.join(os.getcwd(), "binaural_output")
        
        # Create output directory
        os.makedirs(self.output_dir, exist_ok=True)

# Create a default configuration
config = BiAuralConfig()

# You can customize the configuration here
config.duration = 10.0  # 10-second audio
config.movement_type = BiAuralConfig.MovementType.CIRCULAR  # Circular movement around the head
config.room_type = BiAuralConfig.RoomType.SPHERICAL  # Ensure we're using a spherical room
config.room_radius = 2.0  # 2-meter radius spherical room
config.source_type = "pink"  # Pink noise as sound source

print(f"Configured for {config.duration}s of audio with {config.movement_type.value} movement")
print(f"Using a spherical room with radius {config.room_radius}m")
```

```python
class HRTFManager:
    """Handles HRTF loading and interpolation for binaural rendering"""
    
    def __init__(self, config):
        self.config = config
        self.hrir_l = None  # Left ear impulse responses
        self.hrir_r = None  # Right ear impulse responses
        self.azimuths = None  # Azimuth angles for each HRIR pair
        self.elevations = None  # Elevation angles for each HRIR pair
        self.fs_original = None  # Original sampling rate of the HRTF
        
        # Generate synthetic HRTF (for simplicity in this notebook)
        self._generate_synthetic_hrtf()
        
    def _generate_synthetic_hrtf(self):
        """Generate a synthetic HRTF based on a spherical head model"""
        print("Generating synthetic HRTF based on spherical head model")
        
        # Generate a spherical sampling of directions
        n_azimuths = 72   # 5-degree steps in azimuth
        n_elevations = 19  # 10-degree steps in elevation
        
        azimuths = []
        elevations = []
        
        # Create a grid of azimuth/elevation pairs
        for elev_idx in range(n_elevations):
            elev = -90 + elev_idx * 10
            for azim_idx in range(n_azimuths):
                azim = azim_idx * (360 / n_azimuths)
                azimuths.append(azim)
                elevations.append(elev)
        
        self.azimuths = np.array(azimuths)
        self.elevations = np.array(elevations)
        self.distances = np.ones_like(self.azimuths) * 1.0  # Assume 1m distance
        
        # Sample rate for synthetic HRTF
        self.fs_original = self.config.fs
        
        # Create a 512-sample HRIR for each direction
        hrir_length = 512
        self.hrir_l = np.zeros((len(self.azimuths), hrir_length))
        self.hrir_r = np.zeros((len(self.azimuths), hrir_length))
        
        # Generate synthetic HRIR for each direction
        for i, (azim, elev) in enumerate(zip(self.azimuths, self.elevations)):
            self.hrir_l[i], self.hrir_r[i] = self._generate_hrir(azim, elev, hrir_length)
            
        print(f"Generated synthetic HRTF with {len(self.azimuths)} directions")
    
    def _generate_hrir(self, azimuth, elevation, length):
        """Generate a synthetic HRIR for a specific direction based on spherical head model"""
        # Convert to radians
        azimuth_rad = np.radians(azimuth)
        elevation_rad = np.radians(elevation)
        
        # Create empty impulse responses
        hrir_l = np.zeros(length)
        hrir_r = np.zeros(length)
        
        # Center position for the impulse
        center_idx = length // 8
        
        # Calculate ITD (interaural time difference) based on azimuth
        # Using Woodworth's formula for a spherical head
        head_radius = self.config.head_radius
        ear_angle = np.pi/2  # Ear position at 90 degrees
        
        # Calculate the "extra" path length around the head to the contralateral ear
        if -np.pi/2 <= azimuth_rad <= np.pi/2:  # Source on right side or front
            path_diff = head_radius * (azimuth_rad + np.sin(azimuth_rad))
        else:  # Source on left side
            path_diff = head_radius * (np.pi - azimuth_rad + np.sin(np.pi - azimuth_rad))
            
        # Convert path difference to time delay
        itd_time = path_diff / self.config.speed_of_sound
        itd_samples = int(itd_time * self.fs_original)
        
        # Calculate ILD (interaural level difference) - approximate with head shadow
        # More shadow when sound comes from the opposite side
        min_level = 0.1  # Minimum level factor (maximum shadow)
        
        # Calculate level factors for each ear based on azimuth
        # Simple model: level decreases as source moves to opposite side
        if azimuth_rad >= 0:  # Source on right side
            level_r = 1.0
            # Gradually reduce level as azimuth increases (more shadow)
            level_l = 1.0 - (1.0 - min_level) * np.abs(np.sin(azimuth_rad))
        else:  # Source on left side
            level_l = 1.0
            level_r = 1.0 - (1.0 - min_level) * np.abs(np.sin(azimuth_rad))
        
        # Elevation effects - simple pinna model
        # High-frequency boost for frontal sounds, gradual rolloff as elevation changes
        f_cutoff_front = 0.6  # Higher cutoff for frontal sounds
        f_cutoff_above = 0.4  # Medium cutoff for elevated sounds
        f_cutoff_below = 0.2  # Lower cutoff for sounds from below
        
        if -30 <= elevation <= 30:  # Near horizontal plane
            f_cutoff = f_cutoff_front
        elif elevation > 30:  # Above
            f_cutoff = f_cutoff_above
        else:  # Below
            f_cutoff = f_cutoff_below
        
        # Place impulses with appropriate delay
        if azimuth_rad >= 0:  # Source on right side
            # Right ear gets signal first
            hrir_r[center_idx] = level_r
            # Left ear gets delayed signal
            left_idx = center_idx + itd_samples
            if left_idx < length:
                hrir_l[left_idx] = level_l
        else:  # Source on left side
            # Left ear gets signal first
            hrir_l[center_idx] = level_l
            # Right ear gets delayed signal
            right_idx = center_idx + itd_samples
            if right_idx < length:
                hrir_r[right_idx] = level_r
        
        # Apply direction-dependent filtering based on elevation
        b, a = signal.butter(4, f_cutoff, 'low')
        hrir_l = signal.lfilter(b, a, hrir_l)
        hrir_r = signal.lfilter(b, a, hrir_r)
        
        # Add a subtle early reflection to make it sound more natural
        early_delay = int(0.002 * self.fs_original)  # 2ms delay
        reflection_gain = 0.3
        
        if center_idx + early_delay < length:
            hrir_l[center_idx + early_delay] += level_l * reflection_gain
            hrir_r[center_idx + early_delay] += level_r * reflection_gain
        
        return hrir_l, hrir_r
    
    def get_hrir(self, azimuth, elevation, distance=1.0):
        """
        Get HRIR for a specific direction using bilinear interpolation
        
        Parameters:
            azimuth: Azimuth angle in degrees (0=front, 90=left, 180=back, 270=right)
            elevation: Elevation angle in degrees (-90=down, 0=horizontal, 90=up)
            distance: Distance in meters (for near-field compensation)
            
        Returns:
            hrir_l, hrir_r: Left and right ear impulse responses
        """
        # Normalize azimuth to [0, 360) range
        azimuth = azimuth % 360
        
        # Clamp elevation to [-90, 90] range
        elevation = np.clip(elevation, -90, 90)
        
        # Get interpolated HRIR
        if self.config.hrtf_interpolation == BiAuralConfig.HRTFInterpolation.NEAREST:
            hrir_l, hrir_r = self._get_nearest_hrir(azimuth, elevation)
        else:  # Default to bilinear
            hrir_l, hrir_r = self._get_bilinear_hrir(azimuth, elevation)
            
        # Apply near-field compensation if enabled and distance is close
        if self.config.enable_near_field and distance < 1.0:
            hrir_l, hrir_r = self._apply_near_field_compensation(hrir_l, hrir_r, azimuth, distance)
            
        return hrir_l, hrir_r
    
    def _get_nearest_hrir(self, azimuth, elevation):
        """Get HRIR using nearest neighbor interpolation"""
        # Calculate angular distance to each HRTF direction
        azim_diff = np.abs(self.azimuths - azimuth)
        # Handle wrap-around for azimuth
        azim_diff = np.minimum(azim_diff, 360 - azim_diff)
        
        elev_diff = np.abs(self.elevations - elevation)
        
        # Combine differences (weighted to account for azimuth being more sensitive)
        total_diff = azim_diff + 2 * elev_diff
        
        # Find index of minimum difference
        nearest_idx = np.argmin(total_diff)
        
        return self.hrir_l[nearest_idx], self.hrir_r[nearest_idx]
        
    def _get_bilinear_hrir(self, azimuth, elevation):
        """Get HRIR using bilinear interpolation between nearest points"""
        # Find the four nearest HRTF directions for interpolation
        
        # First, find all points within a reasonable range
        azim_diff = np.abs(self.azimuths - azimuth)
        # Handle wrap-around for azimuth
        azim_diff = np.minimum(azim_diff, 360 - azim_diff)
        elev_diff = np.abs(self.elevations - elevation)
        
        # Get points where both azimuth and elevation are close
        close_points = (azim_diff < 45) & (elev_diff < 45)
        
        if np.sum(close_points) < 4:
            # Not enough points for interpolation, fall back to nearest
            return self._get_nearest_hrir(azimuth, elevation)
            
        # Get indices of close points
        close_indices = np.where(close_points)[0]
        
        # Get the coordinates of these points
        close_azimuths = self.azimuths[close_indices]
        close_elevations = self.elevations[close_indices]
        
        # Find points in each quadrant relative to the target direction
        # (higher/lower azimuth, higher/lower elevation)
        q1 = (close_azimuths >= azimuth) & (close_elevations >= elevation)
        q2 = (close_azimuths < azimuth) & (close_elevations >= elevation)
        q3 = (close_azimuths < azimuth) & (close_elevations < elevation)
        q4 = (close_azimuths >= azimuth) & (close_elevations < elevation)
        
        # For each quadrant, find the nearest point
        idx = np.zeros(4, dtype=int)
        
        for i, quad in enumerate([q1, q2, q3, q4]):
            if np.any(quad):
                quad_indices = close_indices[quad]
                quad_azim_diff = azim_diff[quad_indices]
                quad_elev_diff = elev_diff[quad_indices]
                quad_total_diff = quad_azim_diff + 2 * quad_elev_diff
                idx[i] = quad_indices[np.argmin(quad_total_diff)]
            else:
                # If no point in this quadrant, use the overall nearest
                idx[i] = close_indices[np.argmin(azim_diff[close_indices] + 2 * elev_diff[close_indices])]
        
        # Get the coordinates of the four points
        az = self.azimuths[idx]
        el = self.elevations[idx]
        
        # Calculate interpolation weights
        # Normalize coordinates for interpolation
        azim_norm = (azimuth - np.min(az)) / (np.max(az) - np.min(az)) if np.max(az) > np.min(az) else 0.5
        elev_norm = (elevation - np.min(el)) / (np.max(el) - np.min(el)) if np.max(el) > np.min(el) else 0.5
        
        # Bilinear weights
        w = np.zeros(4)
        w[0] = azim_norm * elev_norm
        w[1] = (1 - azim_norm) * elev_norm
        w[2] = (1 - azim_norm) * (1 - elev_norm)
        w[3] = azim_norm * (1 - elev_norm)
        
        # Normalize weights
        w = w / np.sum(w)
        
        # Interpolate HRIRs
        hrir_l = np.zeros_like(self.hrir_l[0])
        hrir_r = np.zeros_like(self.hrir_r[0])
        
        for i in range(4):
            hrir_l += w[i] * self.hrir_l[idx[i]]
            hrir_r += w[i] * self.hrir_r[idx[i]]
            
        return hrir_l, hrir_r
    
    def _apply_near_field_compensation(self, hrir_l, hrir_r, azimuth, distance):
        """Apply compensation for near-field effects"""
        # Only apply if distance is less than 1 meter
        if distance >= 1.0:
            return hrir_l, hrir_r
            
        # Determine which ear is ipsilateral (closer) based on azimuth
        if 0 <= azimuth < 180:
            # Source on the left side
            ipsi_hrir = hrir_l
            contra_hrir = hrir_r
            ipsi_idx = 0
            contra_idx = 1
        else:
            # Source on the right side
            ipsi_hrir = hrir_r
            contra_hrir = hrir_l
            ipsi_idx = 1
            contra_idx = 0
            
        # Calculate near-field adjustments
        # 1. Amplitude boost (inverse square law)
        reference_distance = 1.0  # Standard HRTF measurement distance
        amplitude_boost = (reference_distance / distance) ** 2
        
        # 2. Low-frequency boost based on distance
        # Closer sources have stronger bass response
        boost_freq = 500 + 500 * distance  # Boost frequencies below this value
        boost_freq_norm = boost_freq / (self.config.fs / 2)  # Normalize to [0, 1]
        
        # Apply low shelf filter for bass boost
        b, a = signal.butter(2, boost_freq_norm, btype='lowshelf', analog=False)
        ipsi_boosted = signal.filtfilt(b, a, ipsi_hrir) * amplitude_boost
        
        # 3. Contra-lateral ear has different adjustments
        # Less overall boost but still some
        contra_amplitude_boost = (reference_distance / distance) ** 1.5  # Less than ipsilateral
        
        # Apply same filter but with less boost
        contra_boosted = signal.filtfilt(b, a, contra_hrir) * contra_amplitude_boost
        
        # Return the modified HRIRs in the correct order
        if ipsi_idx == 0:
            return ipsi_boosted, contra_boosted
        else:
            return contra_boosted, ipsi_boosted

# Create HRTF manager with our configuration
hrtf_manager = HRTFManager(config)

# Test by getting HRIRs for a few directions
test_azimuths = [0, 90, 180, 270]
test_elevations = [0, 30, -30]

# Plot a few example HRIRs
plt.figure(figsize=(15, 10))
for i, azim in enumerate(test_azimuths):
    for j, elev in enumerate(test_elevations):
        hrir_l, hrir_r = hrtf_manager.get_hrir(azim, elev)
        
        plt.subplot(len(test_elevations), len(test_azimuths), i + j*len(test_azimuths) + 1)
        plt.plot(hrir_l, 'b-', label='Left ear')
        plt.plot(hrir_r, 'r-', label='Right ear')
        plt.title(f"Az: {azim}°, El: {elev}°")
        if i == 0 and j == 0:
            plt.legend()
        plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
g to reflection
                if refl_dir[0] < 0:  # Reflection coming from left
                    binaural_signal[refl_idx_left, 0] += source_signal[i] * refl_amp_left
                    # Apply head shadow to right ear
                    shadow_factor = 0.7  # Simplified head shadow for reflection
                    binaural_signal[refl_idx_right, 1] += source_signal[i] * refl_amp_right * shadow_factor
                else:  # Reflection coming from right
                    # Apply head shadow to left ear
                    shadow_factor = 0.7  # Simplified head shadow for reflection
                    binaural_signal[refl_idx_left, 0] += source_signal[i] * refl_amp_left * shadow_factor
                    binaural_signal[refl_idx_right, 1] += source_signal[i] * refl_amp_right
    
    # Normalize output
    if np.max(np.abs(binaural_signal)) > 0:
        binaural_signal = binaural_signal / np.max(np.abs(binaural_signal)) * 0.9
    
    return binaural_signal, x_positions


def create_binaural_full(source_signal, direction):
    """Create full binaural sound with all effects including reverb"""
    # First, get the signal with reflections
    binaural_signal, x_positions = create_binaural_with_reflections(source_signal, direction)
    
    # Now add reverb tail
    # Estimate RT60 (reverberation time) based on room properties
    # Simplified Sabine formula: RT60 = 0.161 * V / (A * S)
    # where V = volume, A = absorption coefficient, S = surface area
    room_volume = (4/3) * np.pi * room_radius**3  # Volume of sphere
    room_surface = 4 * np.pi * room_radius**2    # Surface area of sphere
    rt60 = 0.161 * room_volume / (wall_absorption * room_surface)
    
    # Create exponential decay for reverb tail
    reverb_length = int(rt60 * fs)
    
    # Apply reverb as exponential decay
    reverb_gain = 0.3  # Adjust reverb level
    
    # Create decay envelope
    decay_envelope = np.exp(-6.9 * np.linspace(0, 1, reverb_length))
    
    # Apply some frequency-dependent reverb (more high freq absorption)
    # Create two reverb IRs: one for mid/low frequencies, one for high frequencies
    # High frequencies decay faster in real rooms
    decay_env_high = np.exp(-9.0 * np.linspace(0, 1, reverb_length))
    
    # Split signal into two frequency bands
    crossover_freq = 3000  # Crossover frequency in Hz
    crossover_norm = crossover_freq / (fs/2)
    
    # Create filters
    b_low, a_low = signal.butter(2, crossover_norm, 'low')
    b_high, a_high = signal.butter(2, crossover_norm, 'high')
    
    # First, split direct sound + early reflections
    binaural_low = np.zeros_like(binaural_signal)
    binaural_high = np.zeros_like(binaural_signal)
    
    binaural_low[:, 0] = signal.lfilter(b_low, a_low, binaural_signal[:, 0])
    binaural_low[:, 1] = signal.lfilter(b_low, a_low, binaural_signal[:, 1])
    
    binaural_high[:, 0] = signal.lfilter(b_high, a_high, binaural_signal[:, 0])
    binaural_high[:, 1] = signal.lfilter(b_high, a_high, binaural_signal[:, 1])
    
    # Create output buffer with room for reverb tail
    output_length = len(binaural_signal) + reverb_length
    output_signal = np.zeros((output_length, 2))
    
    # Copy direct sound and early reflections
    output_signal[:len(binaural_signal)] = binaural_signal
    
    # Add reverb by convolving the signal with the decay envelopes
    # (simplified approach - true reverb would use convolution with a room impulse response)
    
    # Create random noise bursts for left and right channels
    # (decorrelated for better spatial impression)
    np.random.seed(42)  # For reproducibility
    reverb_noise_l = np.random.randn(reverb_length)
    reverb_noise_r = np.random.randn(reverb_length)
    
    # Low-pass filter the noise to color it like real reverb
    b, a = signal.butter(2, 0.3)  # Low-pass around 6.6 kHz at 44.1 kHz fs
    reverb_noise_l = signal.lfilter(b, a, reverb_noise_l)
    reverb_noise_r = signal.lfilter(b, a, reverb_noise_r)
    
    # Apply decay envelopes differently to different frequency bands
    # Process in small blocks for memory efficiency
    block_size = 1000
    num_blocks = len(binaural_signal) // block_size
    
    for block in range(num_blocks):
        start_idx = block * block_size
        end_idx = min((block + 1) * block_size, len(binaural_signal))
        
        # Create reverb "echogram" by applying energy from this block to the decay
        for i in range(start_idx, end_idx):
            energy_l = (binaural_low[i, 0]**2 + binaural_high[i, 0]**2) ** 0.5
            energy_r = (binaural_low[i, 1]**2 + binaural_high[i, 1]**2) ** 0.5
            
            # Apply different decay rates to different frequency bands
# Low frequencies
            output_signal[i:i+reverb_length, 0] += binaural_low[i, 0] * reverb_noise_l * decay_envelope * reverb_gain
            output_signal[i:i+reverb_length, 1] += binaural_low[i, 1] * reverb_noise_r * decay_envelope * reverb_gain
            
            # High frequencies (faster decay)
            output_signal[i:i+reverb_length, 0] += binaural_high[i, 0] * reverb_noise_l * decay_env_high * reverb_gain
            output_signal[i:i+reverb_length, 1] += binaural_high[i, 1] * reverb_noise_r * decay_env_high * reverb_gain
    
    # Normalize output
    if np.max(np.abs(output_signal)) > 0:
        output_signal = output_signal / np.max(np.abs(output_signal)) * 0.9
    
    return output_signal, x_positions


def main():
    """Generate all binaural audio files"""
    print("Generating source signal...")
    source_signal = generate_pink_noise(duration, fs)
    
    # Create all versions for left-to-right movement
    print("\nProcessing left-to-right movement...")
    
    print("1. Creating basic amplitude panning version...")
    basic_lr, x_pos_lr = create_binaural_basic(source_signal, "left_to_right")
    sf.write(os.path.join(output_dir, "01_basic_left_to_right.wav"), basic_lr, fs)
    
    print("2. Adding interaural time differences...")
    itd_lr, _ = create_binaural_with_itd(source_signal, "left_to_right")
    sf.write(os.path.join(output_dir, "02_with_itd_left_to_right.wav"), itd_lr, fs)
    
    print("3. Adding head shadow effect...")
    shadow_lr, _ = create_binaural_with_head_shadow(source_signal, "left_to_right")
    sf.write(os.path.join(output_dir, "03_with_head_shadow_left_to_right.wav"), shadow_lr, fs)
    
    print("4. Adding distance-based attenuation...")
    distance_lr, _ = create_binaural_with_distance(source_signal, "left_to_right")
    sf.write(os.path.join(output_dir, "04_with_distance_left_to_right.wav"), distance_lr, fs)
    
    print("5. Adding early reflections...")
    reflection_lr, _ = create_binaural_with_reflections(source_signal, "left_to_right")
    sf.write(os.path.join(output_dir, "05_with_reflections_left_to_right.wav"), reflection_lr, fs)
    
    print("6. Adding reverb tail...")
    full_lr, _ = create_binaural_full(source_signal, "left_to_right")
    sf.write(os.path.join(output_dir, "06_full_left_to_right.wav"), full_lr, fs)
    
    # Create all versions for right-to-left movement
    print("\nProcessing right-to-left movement...")
    
    print("1. Creating basic amplitude panning version...")
    basic_rl, x_pos_rl = create_binaural_basic(source_signal, "right_to_left")
    sf.write(os.path.join(output_dir, "01_basic_right_to_left.wav"), basic_rl, fs)
    
    print("2. Adding interaural time differences...")
    itd_rl, _ = create_binaural_with_itd(source_signal, "right_to_left")
    sf.write(os.path.join(output_dir, "02_with_itd_right_to_left.wav"), itd_rl, fs)
    
    print("3. Adding head shadow effect...")
    shadow_rl, _ = create_binaural_with_head_shadow(source_signal, "right_to_left")
    sf.write(os.path.join(output_dir, "03_with_head_shadow_right_to_left.wav"), shadow_rl, fs)
    
    print("4. Adding distance-based attenuation...")
    distance_rl, _ = create_binaural_with_distance(source_signal, "right_to_left")
    sf.write(os.path.join(output_dir, "04_with_distance_right_to_left.wav"), distance_rl, fs)
    
    print("5. Adding early reflections...")
    reflection_rl, _ = create_binaural_with_reflections(source_signal, "right_to_left")
    sf.write(os.path.join(output_dir, "05_with_reflections_right_to_left.wav"), reflection_rl, fs)
    
    print("6. Adding reverb tail...")
    full_rl, _ = create_binaural_full(source_signal, "right_to_left")
    sf.write(os.path.join(output_dir, "06_full_right_to_left.wav"), full_rl, fs)
    
    # Create visualization
    print("\nCreating visualization...")
    plt.figure(figsize=(15, 10))
    
    # Source position
    plt.subplot(2, 2, 1)
    plt.plot(x_pos_lr)
    plt.title("Left-to-Right Movement Path")
    plt.xlabel("Sample")
    plt.ylabel("X Position (m)")
    plt.grid(True)
    
    # Comparison of basic vs full processing for left-to-right
    plt.subplot(2, 2, 2)
    plt.plot(basic_lr[:, 0], 'b-', alpha=0.5, label="Basic Left")
    plt.plot(basic_lr[:, 1], 'r-', alpha=0.5, label="Basic Right")
    plt.plot(full_lr[:, 0], 'b-', alpha=0.8, label="Full Left")
    plt.plot(full_lr[:, 1], 'r-', alpha=0.8, label="Full Right")
    plt.title("Waveform Comparison: Basic vs Full (Left-to-Right)")
    plt.legend()
    plt.grid(True)
    
    # Source position
    plt.subplot(2, 2, 3)
    plt.plot(x_pos_rl)
    plt.title("Right-to-Left Movement Path")
    plt.xlabel("Sample")
    plt.ylabel("X Position (m)")
    plt.grid(True)
    
    # Comparison of basic vs full processing for right-to-left
    plt.subplot(2, 2, 4)
    plt.plot(basic_rl[:, 0], 'b-', alpha=0.5, label="Basic Left")
    plt.plot(basic_rl[:, 1], 'r-', alpha=0.5, label="Basic Right")
    plt.plot(full_rl[:, 0], 'b-', alpha=0.8, label="Full Left")
    plt.plot(full_rl[:, 1], 'r-', alpha=0.8, label="Full Right")
    plt.title("Waveform Comparison: Basic vs Full (Right-to-Left)")
    plt.legend()
    plt.grid(True)
    
    plt.tight_layout()
    plt.savefig(os.path.join(output_dir, "binaural_comparison.png"), dpi=300)
    plt.show()
    
    print(f"\nAll files saved to {output_dir}")
    print("Each movement direction has 6 versions with progressive effects:")
    print("1. Basic amplitude panning only")
    print("2. With interaural time differences (ITD)")
    print("3. With head shadow filtering")
    print("4. With distance-based attenuation")
    print("5. With early reflections")
    print("6. Full version with reverb tail")


if __name__ == "__main__":
    main()
```

```python
"""
Progressive Binaural Effects Generator

This script creates multiple binaural audio files with different levels of realism:
1. Basic: Only amplitude panning (no time delays)
2. With ITD: Adds interaural time differences
3. With Head Shadow: Adds frequency-dependent head shadow effect
4. With Distance: Adds distance-based attenuation
5. With Reflections: Adds early reflections from a spherical room
6. Full: Combines all effects with room reverb

For each effect level, it generates both left-to-right and right-to-left movement files.
"""

import numpy as np
import scipy.signal as signal
import soundfile as sf
import matplotlib.pyplot as plt
import os
from pathlib import Path
import time

# Create output directory
output_dir = os.path.join(os.getcwd(), "binaural_output")
os.makedirs(output_dir, exist_ok=True)

# Basic parameters - Reduced for memory efficiency
fs = 22050                # Sampling frequency (Hz) - reduced from 44100
duration = 2.0            # Duration of sound (seconds) - reduced from 4.0
head_diameter = 0.18      # Diameter of the head (meters)
ear_spacing = 0.18        # Distance between ears (meters)
speed_of_sound = 343.0    # Speed of sound in air (m/s)
room_radius = 2.0         # Radius of spherical room (meters)
wall_absorption = 0.3     # Wall absorption coefficient (0-1)


def generate_pink_noise(duration, fs):
    """Generate pink noise using FFT-based method"""
    # Calculate number of samples
    n_samples = int(duration * fs)
    
    # Generate white noise
    white_noise = np.random.randn(n_samples)
    
    # Get the FFT of the white noise
    X = np.fft.rfft(white_noise)
    
    # Create pink filter (1/f)
    n_bins = len(X)
    f = np.linspace(1, n_bins, n_bins)
    f[0] = 1  # Avoid division by zero
    
    # Apply 1/sqrt(f) filter for power spectrum ~ 1/f
    X = X / np.sqrt(f)
    
    # Convert back to time domain
    signal_out = np.real(np.fft.irfft(X, n_samples))
    
    # Normalize
    signal_out = signal_out / np.max(np.abs(signal_out)) * 0.8
    
    # Apply fade in/out to avoid clicks
    fade_samples = int(0.05 * fs)  # 50ms fade
    fade_in = np.linspace(0, 1, fade_samples)
    fade_out = np.linspace(1, 0, fade_samples)
    
    signal_out[:fade_samples] *= fade_in
    signal_out[-fade_samples:] *= fade_out
    
    return signal_out


def create_source_path(direction, n_samples):
    """Create the source movement path"""
    if direction == "left_to_right":
        start_x = -0.8  # Start 0.8 meters to the left
        end_x = 0.8     # End 0.8 meters to the right
    else:  # right_to_left
        start_x = 0.8   # Start 0.8 meters to the right
        end_x = -0.8    # End 0.8 meters to the left
    
    # Y-coordinate (distance from head center line)
    ordinate_y = 0.2  # 20cm in front of the head at closest approach
    
    # Create position array
    x_positions = np.linspace(start_x, end_x, n_samples)
    y_positions = np.ones(n_samples) * ordinate_y
    
    # Calculate distance from source to head center
    distances_to_center = np.sqrt(x_positions**2 + y_positions**2)
    
    # Calculate coordinates of the ears
    ear_offset = ear_spacing / 2
    left_ear = np.array([-ear_offset, 0])
    right_ear = np.array([ear_offset, 0])
    
    # Calculate distance from source to each ear over time
    distances_to_left = np.zeros(n_samples)
    distances_to_right = np.zeros(n_samples)
    
    for i in range(n_samples):
        source_pos = np.array([x_positions[i], y_positions[i]])
        distances_to_left[i] = np.linalg.norm(source_pos - left_ear)
        distances_to_right[i] = np.linalg.norm(source_pos - right_ear)
    
    return x_positions, y_positions, distances_to_center, distances_to_left, distances_to_right


def apply_head_shadow(source_sample, x_position, i, source_signal):
    """Apply head shadow filtering to a single sample"""
    # Apply head shadow filtering based on current position
    if x_position < 0:  # Source is on the left
        # Apply lowpass filter to right ear (contralateral)
        cutoff_hz = np.interp(abs(x_position), [0, 0.8], [20000, 5000])
        cutoff_norm = cutoff_hz / (fs/2)
        b, a = signal.butter(2, cutoff_norm, 'low')
        
        # Create a short segment around the current sample (for better filtering)
        segment_start = max(0, i - 10)
        segment_end = min(len(source_signal), i + 11)
        segment = source_signal[segment_start:segment_end]
        
        # Apply filter to segment
        filtered_segment = signal.lfilter(b, a, segment)
        
        # Extract the filtered sample
        sample_offset = i - segment_start
        filtered_right = filtered_segment[sample_offset]
        filtered_left = source_sample  # No filter for ipsilateral ear
    else:  # Source is on the right
        # Apply lowpass filter to left ear (contralateral)
        cutoff_hz = np.interp(abs(x_position), [0, 0.8], [20000, 5000])
        cutoff_norm = cutoff_hz / (fs/2)
        b, a = signal.butter(2, cutoff_norm, 'low')
        
        # Process just this sample
        segment_start = max(0, i - 10)
        segment_end = min(len(source_signal), i + 11)
        segment = source_signal[segment_start:segment_end]
        
        # Apply filter to segment
        filtered_segment = signal.lfilter(b, a, segment)
        
        # Extract the filtered sample
        sample_offset = i - segment_start
        filtered_left = filtered_segment[sample_offset]
        filtered_right = source_sample  # No filter for ipsilateral ear
    
    return filtered_left, filtered_right


def create_binaural_basic(source_signal, direction):
    """Create binaural sound with basic amplitude panning only (no time delays)"""
    n_samples = len(source_signal)
    
    # Get source path
    x_positions, _, _, distances_to_left, distances_to_right = create_source_path(direction, n_samples)
    
    # Create simple amplitude panning based on x-position
    # Normalize x to range [-1, 1] where -1 is full left, 1 is full right
    normalized_x = x_positions / np.max(np.abs(x_positions))
    
    # Simple linear panning law
    gain_left = 0.5 - normalized_x * 0.5   # 1.0 at full left, 0.0 at full right
    gain_right = 0.5 + normalized_x * 0.5  # 0.0 at full left, 1.0 at full right
    
    # Apply panning to create stereo output
    binaural_signal = np.zeros((n_samples, 2))
    binaural_signal[:, 0] = source_signal * gain_left
    binaural_signal[:, 1] = source_signal * gain_right
    
    return binaural_signal, x_positions


def create_binaural_with_itd(source_signal, direction):
    """Create binaural sound with amplitude panning and interaural time differences"""
    n_samples = len(source_signal)
    
    # Get source path
    x_positions, _, _, distances_to_left, distances_to_right = create_source_path(direction, n_samples)
    
    # Calculate delays in samples
    delays_left = np.round(distances_to_left / speed_of_sound * fs).astype(int)
    delays_right = np.round(distances_to_right / speed_of_sound * fs).astype(int)
    
    # Normalize delays to have minimum delay of 0
    min_delay = min(np.min(delays_left), np.min(delays_right))
    delays_left -= min_delay
    delays_right -= min_delay
    
    # Find maximum delay to determine output buffer size
    max_delay = max(np.max(delays_left), np.max(delays_right))
    
    # Create output buffer
    output_length = n_samples + max_delay
    binaural_signal = np.zeros((output_length, 2))
    
    # Create simple amplitude panning based on x-position
    normalized_x = x_positions / np.max(np.abs(x_positions))
    gain_left = 0.5 - normalized_x * 0.5
    gain_right = 0.5 + normalized_x * 0.5
    
    # Process each sample with its calculated delay and amplitude
    for i in range(n_samples):
        left_idx = i + delays_left[i]
        right_idx = i + delays_right[i]
        
        if left_idx < output_length:
            binaural_signal[left_idx, 0] += source_signal[i] * gain_left[i]
        
        if right_idx < output_length:
            binaural_signal[right_idx, 1] += source_signal[i] * gain_right[i]
    
    # Normalize output
    if np.max(np.abs(binaural_signal)) > 0:
        binaural_signal = binaural_signal / np.max(np.abs(binaural_signal)) * 0.9
    
    return binaural_signal, x_positions


def create_binaural_with_head_shadow(source_signal, direction):
    """Create binaural sound with ITD and frequency-dependent head shadow effect"""
    n_samples = len(source_signal)
    
    # Get source path
    x_positions, _, _, distances_to_left, distances_to_right = create_source_path(direction, n_samples)
    
    # Calculate delays in samples
    delays_left = np.round(distances_to_left / speed_of_sound * fs).astype(int)
    delays_right = np.round(distances_to_right / speed_of_sound * fs).astype(int)
    
    # Normalize delays to have minimum delay of 0
    min_delay = min(np.min(delays_left), np.min(delays_right))
    delays_left -= min_delay
    delays_right -= min_delay
    
    # Find maximum delay to determine output buffer size
    max_delay = max(np.max(delays_left), np.max(delays_right))
    
    # Create output buffer
    output_length = n_samples + max_delay
    binaural_signal = np.zeros((output_length, 2))
    
    # Create amplitude panning based on x-position
    normalized_x = x_positions / np.max(np.abs(x_positions))
    gain_left = 0.5 - normalized_x * 0.5
    gain_right = 0.5 + normalized_x * 0.5
    
    # Process each sample with filtering applied on-the-fly
    for i in range(n_samples):
        # Apply head shadow filtering
        filtered_left, filtered_right = apply_head_shadow(source_signal[i], x_positions[i], i, source_signal)
        
        # Add to output with delay
        left_idx = i + delays_left[i]
        right_idx = i + delays_right[i]
        
        if left_idx < output_length:
            binaural_signal[left_idx, 0] += filtered_left * gain_left[i]
        
        if right_idx < output_length:
            binaural_signal[right_idx, 1] += filtered_right * gain_right[i]
    
    # Normalize output
    if np.max(np.abs(binaural_signal)) > 0:
        binaural_signal = binaural_signal / np.max(np.abs(binaural_signal)) * 0.9
    
    return binaural_signal, x_positions


def create_binaural_with_distance(source_signal, direction):
    """Create binaural sound with ITD, head shadow, and distance-based attenuation"""
    n_samples = len(source_signal)
    
    # Get source path
    x_positions, _, distances_to_center, distances_to_left, distances_to_right = create_source_path(direction, n_samples)
    
    # Calculate delays in samples
    delays_left = np.round(distances_to_left / speed_of_sound * fs).astype(int)
    delays_right = np.round(distances_to_right / speed_of_sound * fs).astype(int)
    
    # Normalize delays to have minimum delay of 0
    min_delay = min(np.min(delays_left), np.min(delays_right))
    delays_left -= min_delay
    delays_right -= min_delay
    
    # Find maximum delay to determine output buffer size
    max_delay = max(np.max(delays_left), np.max(delays_right))
    
    # Create output buffer
    output_length = n_samples + max_delay
    binaural_signal = np.zeros((output_length, 2))
    
    # Calculate distance-based attenuation (inverse square law)
    att_factor = 40  # Attenuation factor (adjust as needed)
    
    # Calculate amplitudes with distance attenuation
    # Baseline amplitude: inverse square law based on distance to head center
    base_amplitude = 1.0 / (distances_to_center ** 2) * att_factor
    # Normalize to avoid very loud sounds when source is close
    base_amplitude = base_amplitude / np.max(base_amplitude) * 0.9
    
    # Apply head position offset to create stereo amplitude
    normalized_x = x_positions / np.max(np.abs(x_positions))
    pan_left = 0.5 - normalized_x * 0.5
    pan_right = 0.5 + normalized_x * 0.5
    
    gain_left = base_amplitude * pan_left
    gain_right = base_amplitude * pan_right
    
    # Process each sample with filtering applied on-the-fly
    for i in range(n_samples):
        # Apply head shadow filtering
        filtered_left, filtered_right = apply_head_shadow(source_signal[i], x_positions[i], i, source_signal)
        
        # Add to output with delay
        left_idx = i + delays_left[i]
        right_idx = i + delays_right[i]
        
        if left_idx < output_length:
            binaural_signal[left_idx, 0] += filtered_left * gain_left[i]
        
        if right_idx < output_length:
            binaural_signal[right_idx, 1] += filtered_right * gain_right[i]
    
    # Normalize output
    if np.max(np.abs(binaural_signal)) > 0:
        binaural_signal = binaural_signal / np.max(np.abs(binaural_signal)) * 0.9
    
    return binaural_signal, x_positions


def create_binaural_with_reflections(source_signal, direction):
    """Create binaural sound with ITD, head shadow, distance attenuation, and early reflections"""
    n_samples = len(source_signal)
    
    # Get source path
    x_positions, y_positions, distances_to_center, distances_to_left, distances_to_right = create_source_path(direction, n_samples)
    
    # Calculate delays in samples
    delays_left = np.round(distances_to_left / speed_of_sound * fs).astype(int)
    delays_right = np.round(distances_to_right / speed_of_sound * fs).astype(int)
    
    # Normalize delays to have minimum delay of 0
    min_delay = min(np.min(delays_left), np.min(delays_right))
    delays_left -= min_delay
    delays_right -= min_delay
    
    # Calculate maximum reflection delay (based on room size)
    max_reflection_distance = 2 * room_radius  # Maximum possible path: source->wall->listener
    max_reflection_delay = int(max_reflection_distance / speed_of_sound * fs)
    
    # Find maximum delay to determine output buffer size
    max_delay = max(np.max(delays_left), np.max(delays_right)) + max_reflection_delay
    
    # Create output buffer
    output_length = n_samples + max_delay
    binaural_signal = np.zeros((output_length, 2))
    
    # Calculate distance-based attenuation
    att_factor = 40
    base_amplitude = 1.0 / (distances_to_center ** 2) * att_factor
    base_amplitude = base_amplitude / np.max(base_amplitude) * 0.9
    
    # Apply head position offset to create stereo amplitude
    normalized_x = x_positions / np.max(np.abs(x_positions))
    pan_left = 0.5 - normalized_x * 0.5
    pan_right = 0.5 + normalized_x * 0.5
    
    gain_left = base_amplitude * pan_left
    gain_right = base_amplitude * pan_right
    
    # First, add direct sound (with on-the-fly filtering)
    for i in range(n_samples):
        # Apply head shadow filtering
        filtered_left, filtered_right = apply_head_shadow(source_signal[i], x_positions[i], i, source_signal)
        
        left_idx = i + delays_left[i]
        right_idx = i + delays_right[i]
        
        if left_idx < output_length:
            binaural_signal[left_idx, 0] += filtered_left * gain_left[i]
        
        if right_idx < output_length:
            binaural_signal[right_idx, 1] += filtered_right * gain_right[i]
    
    # Now add early reflections
    # We'll use 6 virtual sources (like reflections from 6 points on the spherical room)
    reflection_points = [
        [0, 0, room_radius],   # Ceiling
        [0, 0, -room_radius],  # Floor
        [room_radius, 0, 0],   # Right wall
        [-room_radius, 0, 0],  # Left wall
        [0, room_radius, 0],   # Back wall
        [0, -room_radius, 0]   # Front wall
    ]
    
    # Reflection coefficient (how much sound is reflected vs. absorbed)
    reflection_coef = 1.0 - wall_absorption
    
    # Calculate and add reflections - increased step size to reduce computation
    step_size = 50  # Process only every 50th sample for reflections
    
    for refl_point in reflection_points:
        refl_point = np.array(refl_point)
        
        # Process each time step with larger steps for efficiency
        for i in range(0, n_samples, step_size):
            # Current source position (2D + zero z-coordinate)
            source_pos = np.array([x_positions[i], y_positions[i], 0])
            
            # Calculate reflection path:
            # 1. Direction from source to reflection point
            dir_to_refl = refl_point - source_pos
            dist_to_refl = np.linalg.norm(dir_to_refl)
            
            # 2. Normalized direction
            norm_dir = dir_to_refl / (dist_to_refl + 1e-10)  # Avoid division by zero
            
            # 3. Calculate reflection point on the sphere
            reflection_hit = room_radius * norm_dir
            
            # 4. Calculate reflected ray direction
            refl_dir = -norm_dir
            
            # 5. Calculate path from reflection to each ear
            left_ear_pos = np.array([-ear_spacing/2, 0, 0])
            right_ear_pos = np.array([ear_spacing/2, 0, 0])
            
            refl_to_left = left_ear_pos - reflection_hit
            refl_to_right = right_ear_pos - reflection_hit
            
            dist_refl_to_left = np.linalg.norm(refl_to_left)
            dist_refl_to_right = np.linalg.norm(refl_to_right)
            
            # 6. Calculate total path length and resulting delay
            total_path_left = dist_to_refl + dist_refl_to_left
            total_path_right = dist_to_refl + dist_refl_to_right
            
            delay_refl_left = int(total_path_left / speed_of_sound * fs)
            delay_refl_right = int(total_path_right / speed_of_sound * fs)
            
            # 7. Calculate reflection amplitude
            refl_amp_left = (reflection_coef / total_path_left) * 0.2
            refl_amp_right = (reflection_coef / total_path_right) * 0.2
            
            # 8. Add reflection to output buffer
            refl_idx_left = i + delay_refl_left
            refl_idx_right = i + delay_refl_right
            
            if refl_idx_left < output_length and i < n_samples:
                # Apply head shadow filtering to reflection
                if refl_dir[0] < 0:  # Reflection coming from left
                    binaural_signal[refl_idx_left, 0] += source_signal[i] * refl_amp_left
                    # Apply head shadow to right ear
                    shadow_factor = 0.7  # Simplified head shadow for reflection
                    binaural_signal[refl_idx_right, 1] += source_signal[i] * refl_amp_right * shadow_factor
                else:  # Reflection coming from right
                    # Apply head shadow to left ear
                    shadow_factor = 0.7  # Simplified head shadow for reflection
                    binaural_signal[refl_idx_left, 0] += source_signal[i] * refl_amp_left * shadow_factor
                    binaural_signal[refl_idx_right, 1] += source_signal[i] * refl_amp_right
    
    # Normalize output
    if np.max(np.abs(binaural_signal)) > 0:
        binaural_signal = binaural_signal / np.max(np.abs(binaural_signal)) * 0.9
    
    return binaural_signal, x_positions


def create_binaural_full(source_signal, direction):
    """Create full binaural sound with all effects including reverb"""
    # First, get the signal with reflections
    binaural_signal, x_positions = create_binaural_with_reflections(source_signal, direction)
    
    # Now add reverb tail
    # Estimate RT60 based on room properties
    room_volume = (4/3) * np.pi * room_radius**3  # Volume of sphere
    room_surface = 4 * np.pi * room_radius**2    # Surface area of sphere
    rt60 = 0.161 * room_volume / (wall_absorption * room_surface)
    
    # Create shorter reverb length to save memory
    reverb_length = min(int(rt60 * fs), 22050)  # Cap at 1 second
    
    # Apply reverb as exponential decay
    reverb_gain = 0.3  # Adjust reverb level
    
    # Create decay envelope
    decay_envelope = np.exp(-6.9 * np.linspace(0, 1, reverb_length))
    decay_env_high = np.exp(-9.0 * np.linspace(0, 1, reverb_length))
    
    # Create random noise bursts for left and right channels
    np.random.seed(42)  # For reproducibility
    reverb_noise_l = np.random.randn(reverb_length)
    reverb_noise_r = np.random.randn(reverb_length)
    
    # Low-pass filter the noise
    b, a = signal.butter(2, 0.3)  # Low-pass around 6.6 kHz at 44.1 kHz fs
    reverb_noise_l = signal.lfilter(b, a, reverb_noise_l)
    reverb_noise_r = signal.lfilter(b, a, reverb_noise_r)
    
    # Split signal into two frequency bands
    crossover_freq = 3000  # Crossover frequency in Hz
    crossover_norm = crossover_freq / (fs/2)
    b_low, a_low = signal.butter(2, crossover_norm, 'low')
    b_high, a_high = signal.butter(2, crossover_norm, 'high')
    
    # Create output buffer with room for reverb tail
    output_length = len(binaural_signal) + reverb_length
    output_signal = np.zeros((output_length, 2))
    
    # Copy direct sound and early reflections
    output_signal[:len(binaural_signal)] = binaural_signal
    
    # First apply filters to entire signal to get frequency bands
    binaural_low_left = signal.lfilter(b_low, a_low, binaural_signal[:, 0])
    binaural_low_right = signal.lfilter(b_low, a_low, binaural_signal[:, 1])
    binaural_high_left = signal.lfilter(b_high, a_high, binaural_signal[:, 0])
    binaural_high_right = signal.lfilter(b_high, a_high, binaural_signal[:, 1])
    
    # Process in blocks with further increased step size for even better memory efficiency
    block_size = 200  # Smaller block size
    num_blocks = len(binaural_signal) // block_size
    
    for block in range(num_blocks):
        start_idx = block * block_size
        end_idx = min((block + 1) * block_size, len(binaural_signal))
        
        for i in range(start_idx, end_idx, 20):  # Skip samples to reduce computation
            # Apply low freq reverb
            output_signal[i:i+reverb_length, 0] += binaural_low_left[i] * reverb_noise_l * decay_envelope * reverb_gain * 0.5
            output_signal[i:i+reverb_length, 1] += binaural_low_right[i] * reverb_noise_r * decay_envelope * reverb_gain * 0.5
            
            # Apply high freq reverb (faster decay)
            output_signal[i:i+reverb_length, 0] += binaural_high_left[i] * reverb_noise_l * decay_env_high * reverb_gain * 0.5
            output_signal[i:i+reverb_length, 1] += binaural_high_right[i] * reverb_noise_r * decay_env_high * reverb_gain * 0.5
    
    # Normalize output
    if np.max(np.abs(output_signal)) > 0:
        output_signal = output_signal / np.max(np.abs(output_signal)) * 0.9
    
    return output_signal, x_positions


def main():
    """Generate all binaural audio files"""
    print("Generating source signal...")
    source_signal = generate_pink_noise(duration, fs)
    
    # Create all versions for left-to-right movement
    print("\nProcessing left-to-right movement...")
    
    print("1. Creating basic amplitude panning version...")
    basic_lr, x_pos_lr = create_binaural_basic(source_signal, "left_to_right")
    sf.write(os.path.join(output_dir, "01_basic_left_to_right.wav"), basic_lr, fs)
    
    print("2. Adding interaural time differences...")
    itd_lr, _ = create_binaural_with_itd(source_signal, "left_to_right")
    sf.write(os.path.join(output_dir, "02_with_itd_left_to_right.wav"), itd_lr, fs)
    
    print("3. Adding head shadow effect...")
    shadow_lr, _ = create_binaural_with_head_shadow(source_signal, "left_to_right")
    sf.write(os.path.join(output_dir, "03_with_head_shadow_left_to_right.wav"), shadow_lr, fs)
    
    print("4. Adding distance-based attenuation...")
    distance_lr, _ = create_binaural_with_distance(source_signal, "left_to_right")
    sf.write(os.path.join(output_dir, "04_with_distance_left_to_right.wav"), distance_lr, fs)
    
    print("5. Adding early reflections...")
    reflection_lr, _ = create_binaural_with_reflections(source_signal, "left_to_right")
    sf.write(os.path.join(output_dir, "05_with_reflections_left_to_right.wav"), reflection_lr, fs)
    
    print("6. Adding reverb tail...")
    full_lr, _ = create_binaural_full(source_signal, "left_to_right")
    sf.write(os.path.join(output_dir, "06_full_left_to_right.wav"), full_lr, fs)
    
    # Create all versions for right-to-left movement
    print("\nProcessing right-to-left movement...")
    
    print("1. Creating basic amplitude panning version...")
    basic_rl, x_pos_rl = create_binaural_basic(source_signal, "right_to_left")
    sf.write(os.path.join(output_dir, "01_basic_right_to_left.wav"), basic_rl, fs)
    
    print("2. Adding interaural time differences...")
    itd_rl, _ = create_binaural_with_itd(source_signal, "right_to_left")
    sf.write(os.path.join(output_dir, "02_with_itd_right_to_left.wav"), itd_rl, fs)
    
    print("3. Adding head shadow effect...")
    shadow_rl, _ = create_binaural_with_head_shadow(source_signal, "right_to_left")
    sf.write(os.path.join(output_dir, "03_with_head_shadow_right_to_left.wav"), shadow_rl, fs)
    
    print("4. Adding distance-based attenuation...")
    distance_rl, _ = create_binaural_with_distance(source_signal, "right_to_left")
    sf.write(os.path.join(output_dir, "04_with_distance_right_to_left.wav"), distance_rl, fs)
    print("5. Adding early reflections...")
    reflection_rl, _ = create_binaural_with_reflections(source_signal, "right_to_left")
    sf.write(os.path.join(output_dir, "05_with_reflections_right_to_left.wav"), reflection_rl, fs)
    
    print("6. Adding reverb tail...")
    full_rl, _ = create_binaural_full(source_signal, "right_to_left")
    sf.write(os.path.join(output_dir, "06_full_right_to_left.wav"), full_rl, fs)
    
    # Create visualization
    print("\nCreating visualization...")
    plt.figure(figsize=(15, 10))
    
    # Source position
    plt.subplot(2, 2, 1)
    plt.plot(x_pos_lr)
    plt.title("Left-to-Right Movement Path")
    plt.xlabel("Sample")
    plt.ylabel("X Position (m)")
    plt.grid(True)
    
    # Comparison of basic vs full processing for left-to-right
    plt.subplot(2, 2, 2)
    plt.plot(basic_lr[:, 0], 'b-', alpha=0.5, label="Basic Left")
    plt.plot(basic_lr[:, 1], 'r-', alpha=0.5, label="Basic Right")
    plt.plot(full_lr[:, 0], 'b-', alpha=0.8, label="Full Left")
    plt.plot(full_lr[:, 1], 'r-', alpha=0.8, label="Full Right")
    plt.title("Waveform Comparison: Basic vs Full (Left-to-Right)")
    plt.legend()
    plt.grid(True)
    
    # Source position
    plt.subplot(2, 2, 3)
    plt.plot(x_pos_rl)
    plt.title("Right-to-Left Movement Path")
    plt.xlabel("Sample")
    plt.ylabel("X Position (m)")
    plt.grid(True)
    
    # Comparison of basic vs full processing for right-to-left
    plt.subplot(2, 2, 4)
    plt.plot(basic_rl[:, 0], 'b-', alpha=0.5, label="Basic Left")
    plt.plot(basic_rl[:, 1], 'r-', alpha=0.5, label="Basic Right")
    plt.plot(full_rl[:, 0], 'b-', alpha=0.8, label="Full Left")
    plt.plot(full_rl[:, 1], 'r-', alpha=0.8, label="Full Right")
    plt.title("Waveform Comparison: Basic vs Full (Right-to-Left)")
    plt.legend()
    plt.grid(True)
    
    plt.tight_layout()
    plt.savefig(os.path.join(output_dir, "binaural_comparison.png"), dpi=300)
    plt.show()
    
    print(f"\nAll files saved to {output_dir}")
    print("Each movement direction has 6 versions with progressive effects:")
    print("1. Basic amplitude panning only")
    print("2. With interaural time differences (ITD)")
    print("3. With head shadow filtering")
    print("4. With distance-based attenuation")
    print("5. With early reflections")
    print("6. Full version with reverb tail")


if __name__ == "__main__":
    main()
```
