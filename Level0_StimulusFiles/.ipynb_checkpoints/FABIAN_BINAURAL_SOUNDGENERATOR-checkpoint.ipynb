```python
"""
Enhanced SOFA HRTF Distance-Based Movement

This script creates binaural audio files where a sound source approaches the listener
from the left side (-90° azimuth) at ear level (0° elevation), maintaining the same
angle but decreasing in distance, for various HRTF models. 

The script includes room acoustics, looming effect, and Doppler effect to create
a more realistic approaching sound.
"""
import numpy as np
import scipy.signal as signal
import soundfile as sf
import matplotlib.pyplot as plt
import os
import time

# First, ensure required libraries are installed
try:
    import sofar
    SOFA_AVAILABLE = True
except ImportError:
    print("Warning: sofar library not found. Please install with: pip install sofar")
    SOFA_AVAILABLE = False

try:
    import pyroomacoustics as pra
    PYROOMACOUSTICS_AVAILABLE = True
except ImportError:
    print("Warning: pyroomacoustics library not found. Please install with: pip install pyroomacoustics")
    PYROOMACOUSTICS_AVAILABLE = False

#############################################################
# CONTROL PARAMETERS - ADJUST THESE TO CHANGE THE SIMULATION
#############################################################

# Spatial parameters
INITIAL_DISTANCE = 2.0    # Starting distance in meters
FINAL_DISTANCE = 0.3      # Final distance in meters (closest approach)
CONSTANT_AZIMUTH = -90    # Fixed angle (directly to the left of listener)
CONSTANT_ELEVATION = 0    # Fixed elevation (ear level)

# Time parameters
DURATION = 3.0            # Duration of movement in seconds (longer = slower movement)
SAMPLE_RATE = 44100       # Audio sample rate in Hz
N_SEGMENTS = 60           # Number of segments for processing (higher = smoother movement)

# Room acoustics parameters
ROOM_RADIUS = 2.0         # Radius of the spherical room in meters
WALL_ABSORPTION = 0.3     # Wall absorption coefficient (0-1, higher = more absorption)
RT60 = 0.4                # Desired RT60 reverberation time (seconds)

# Effect toggles
ENABLE_ROOM_ACOUSTICS = True   # Whether to simulate room acoustics
ENABLE_DOPPLER = True          # Whether to simulate Doppler effect 
ENABLE_LOOMING = True          # Whether to enhance approaching sound with looming effect

# List of SOFA files to compare
sofa_files = [
    r"C:\Users\cogpsy-vrlab\Downloads\HRIRs_neutral_head_orientation_v4\SOFA\FABIAN_HRIR_measured_HATO_0.sofa",
    r"C:\Users\cogpsy-vrlab\Downloads\HRIRs_neutral_head_orientation_v4\SOFA\FABIAN_HRIR_modeled_HATO_0.sofa",
    r"C:\Users\cogpsy-vrlab\Downloads\HRIRs_neutral_head_orientation_v4\SOFA\FABIAN_CTF_measured.sofa",
    r"C:\Users\cogpsy-vrlab\Downloads\HRIRs_neutral_head_orientation_v4\SOFA\FABIAN_CTF_measured_smoothed.sofa"
]

# Create output directory
output_dir = os.path.join(os.getcwd(), "binaural_approaching_enhanced")
os.makedirs(output_dir, exist_ok=True)

# Calculate speed of sound (m/s)
SPEED_OF_SOUND = 343.0

def generate_pink_noise(duration, fs):
    """Generate pink noise using FFT-based method"""
    n_samples = int(duration * fs)
    
    # Generate white noise
    np.random.seed(42)  # For reproducibility
    white_noise = np.random.randn(n_samples)
    
    # Get the FFT of the white noise
    X = np.fft.rfft(white_noise)
    
    # Create pink filter (1/f)
    n_bins = len(X)
    f = np.linspace(1, n_bins, n_bins)
    f[0] = 1  # Avoid division by zero
    
    # Apply 1/sqrt(f) filter for power spectrum ~ 1/f
    X = X / np.sqrt(f)
    
    # Convert back to time domain
    pink_noise = np.real(np.fft.irfft(X, n_samples))
    
    # Normalize
    pink_noise = pink_noise / np.max(np.abs(pink_noise)) * 0.8
    
    # Apply fade in/out to avoid clicks
    fade_samples = int(0.05 * fs)  # 50ms fade
    fade_in = np.linspace(0, 1, fade_samples)
    fade_out = np.linspace(1, 0, fade_samples)
    
    pink_noise[:fade_samples] *= fade_in
    pink_noise[-fade_samples:] *= fade_out
    
    return pink_noise

def load_sofa_file(sofa_path):
    """Load HRTF data from SOFA file"""
    try:
        print(f"Loading SOFA file: {os.path.basename(sofa_path)}")
        
        # Load the SOFA file
        hrtf = sofar.read_sofa(sofa_path)
        
        # Display basic SOFA info
        print(f"SOFA convention: {hrtf.GLOBAL_SOFAConventions} v{hrtf.GLOBAL_SOFAConventionsVersion}")
        
        # Get dimensions
        dimensions = {}
        for dim_name in ['M', 'R', 'N']:
            try:
                dimensions[dim_name] = hrtf.get_dimension(dim_name)
                print(f"{dim_name} = {dimensions[dim_name]}")
            except:
                dimensions[dim_name] = None
                print(f"{dim_name} = Unknown")
        
        return hrtf, dimensions
        
    except Exception as e:
        print(f"Error loading SOFA file: {e}")
        import traceback
        traceback.print_exc()
        return None, None

def find_nearest_hrtf(hrtf, target_azimuth, target_elevation=0, target_distance=1.0):
    """Find the measurement closest to the target azimuth, elevation, and distance"""
    try:
        source_positions = np.array(hrtf.SourcePosition)
        
        # Handle different position array shapes
        if source_positions.shape[0] == 3 and len(source_positions.shape) == 2:
            source_positions = source_positions.T
        
        # Check if positions are in degrees or radians
        position_type = hrtf.SourcePosition_Type
        max_azimuth = np.max(np.abs(source_positions[:, 0]))
        positions_in_degrees = max_azimuth > 6.28
        
        if positions_in_degrees:
            # Both source positions and target are in degrees
            az_compare = target_azimuth
            el_compare = target_elevation
        else:
            # Source positions in radians, convert target to radians
            az_compare = np.deg2rad(target_azimuth)
            el_compare = np.deg2rad(target_elevation)
        
        # Handle negative azimuths if data is in 0-360 format
        if positions_in_degrees and target_azimuth < 0:
            # Check if we have negative azimuths in the data
            has_negative_az = np.any(source_positions[:, 0] < 0)
            if not has_negative_az:
                # Convert target from -180 to 180 format to 0 to 360 format
                az_compare = (target_azimuth + 360) % 360
        
        # Create distance metric
        # For spherical coordinates: focus mostly on azimuth, then elevation
        # Ignore distance since most HRTF databases only have one distance
        if position_type == 'spherical':
            dists = np.sqrt(
                2 * (source_positions[:, 0] - az_compare)**2 + 
                (source_positions[:, 1] - el_compare)**2
            )
        else:
            # For Cartesian: calculate actual 3D distance
            # Create a target position in Cartesian
            if position_type == 'cartesian':
                # Convert spherical to Cartesian
                target_x = target_distance * np.cos(np.deg2rad(target_elevation)) * np.cos(np.deg2rad(target_azimuth))
                target_y = target_distance * np.cos(np.deg2rad(target_elevation)) * np.sin(np.deg2rad(target_azimuth))
                target_z = target_distance * np.sin(np.deg2rad(target_elevation))
                target_pos = np.array([target_x, target_y, target_z])
                
                # Calculate Euclidean distance to each position
                dists = np.sqrt(np.sum((source_positions - target_pos)**2, axis=1))
            else:
                # Unknown position type, use simple difference
                dists = np.abs(source_positions[:, 0] - az_compare)
                
        # Find nearest position
        nearest_idx = np.argmin(dists)
        nearest_pos = source_positions[nearest_idx]
        
        return nearest_idx, nearest_pos
        
    except Exception as e:
        print(f"Error finding nearest HRTF: {e}")
        import traceback
        traceback.print_exc()
        return 0, None

def extract_hrtf_data(hrtf, idx):
    """Extract HRTF impulse responses for a given index"""
    hrtf_data = hrtf.Data_IR
    
    # Check dimensions and extract
    if hrtf_data.shape[0] == 2 and hrtf_data.shape[1] > 2:
        # Shape is (R, M, N)
        hrtf_left = hrtf_data[0, idx, :]
        hrtf_right = hrtf_data[1, idx, :]
    else:
        # Shape is (M, R, N)
        try:
            hrtf_left = hrtf_data[idx, 0, :]
            hrtf_right = hrtf_data[idx, 1, :]
        except:
            # Try other dimension orderings if the above fails
            try:
                if len(hrtf_data.shape) == 4:  # If 4D array
                    hrtf_left = hrtf_data[idx, 0, 0, :]
                    hrtf_right = hrtf_data[idx, 1, 0, :]
                else:
                    raise ValueError(f"Unexpected HRTF data shape: {hrtf_data.shape}")
            except Exception as e:
                print(f"Error extracting HRTF data: {e}")
                # Return empty IRs as fallback
                dummy_ir = np.zeros(256)
                dummy_ir[0] = 1.0
                return dummy_ir, dummy_ir
    
    return hrtf_left, hrtf_right

def setup_room_model():
    """Create a spherical room model using pyroomacoustics"""
    if not PYROOMACOUSTICS_AVAILABLE or not ENABLE_ROOM_ACOUSTICS:
        print("Room acoustics disabled or pyroomacoustics not available")
        return None
    
    try:
        # Calculate actual RT60 based on room parameters
        # Using Sabine's formula: RT60 = 0.161 * V / (A * S)
        # where V is volume, A is absorption coefficient, S is surface area
        room_volume = (4/3) * np.pi * ROOM_RADIUS**3
        room_surface = 4 * np.pi * ROOM_RADIUS**2
        
        # Adjust absorption to match desired RT60 if possible
        sabine_absorption = 0.161 * room_volume / (RT60 * room_surface)
        
        # Clamp absorption between 0.01 and 0.99
        absorption = min(0.99, max(0.01, sabine_absorption))
        
        print(f"Creating spherical room model with radius {ROOM_RADIUS}m")
        print(f"Calculated absorption coefficient: {absorption:.2f} for RT60 of {RT60}s")
        
        # Create a circular room approximation (2D projection of sphere)
        num_walls = 16
        angles = np.linspace(0, 2*np.pi, num_walls, endpoint=False)
        corners = np.array([[ROOM_RADIUS * np.cos(ang), ROOM_RADIUS * np.sin(ang)] 
                          for ang in angles]).T
        
        # Create room with the specified absorption
        materials = pra.Material(energy_absorption=absorption)
        room = pra.Room.from_corners(corners, fs=SAMPLE_RATE, materials=materials, max_order=3)
        
        # Create stereo microphone array (listener's ears)
        ear_spacing = 0.18  # typical head size
        mic_positions = np.array([
            [-ear_spacing/2, 0],  # Left ear
            [ear_spacing/2, 0]    # Right ear
        ]).T
        
        mic_array = pra.MicrophoneArray(mic_positions, room.fs)
        room.add_microphone_array(mic_array)
        
        return room
    
    except Exception as e:
        print(f"Error setting up room model: {e}")
        import traceback
        traceback.print_exc()
        return None

def apply_room_acoustics(audio, source_position, room):
    """Apply room acoustics to the audio using the pyroomacoustics model"""
    if room is None:
        return audio
    
    try:
        # Reset any existing sources
        room.sources = []
        
        # Add new source with the given position and mono audio
        # For simplicity, use the average of stereo channels
        if audio.ndim == 2:
            mono_signal = np.mean(audio, axis=1)
        else:
            mono_signal = audio
        
        # Add the source to the room
        room.add_source(source_position, signal=mono_signal)
        
        # Compute the room impulse responses
        room.compute_rir()
        
        # Simulate the room acoustics
        room_output = room.simulate()
        
        # Convert to stereo output
        output_signal = np.zeros((len(room_output), 2))
        output_signal[:, 0] = room_output[:, 0]  # Left ear
        output_signal[:, 1] = room_output[:, 1]  # Right ear
        
        # Normalize output
        max_val = np.max(np.abs(output_signal))
        if max_val > 0:
            output_signal = output_signal / max_val * 0.9
            
        return output_signal
        
    except Exception as e:
        print(f"Error applying room acoustics: {e}")
        import traceback
        traceback.print_exc()
        return audio

def apply_basic_distance_model(audio, distance, reference_distance=1.0):
    """
    Apply a basic distance model to simulate source distance
    
    This uses the inverse square law for amplitude and
    adds a low-pass filter effect for further distances
    """
    # Apply inverse square law for amplitude
    distance_ratio = max(0.1, min(distance / reference_distance, 10.0))
    amplitude_scale = 1.0 / (distance_ratio ** 2)
    
    # Normalize scale to avoid extreme amplification
    amplitude_scale = min(amplitude_scale, 3.0)
    distance_audio = audio * amplitude_scale
    
    # Apply distance-based filtering (closer = brighter, further = more muffled)
    if distance > reference_distance:
        # Create a lowpass filter for distant sounds
        cutoff_freq = max(1000, min(20000, 20000 / distance_ratio))
        cutoff_norm = cutoff_freq / (SAMPLE_RATE/2)  # Normalize to Nyquist
        b, a = signal.butter(2, cutoff_norm, btype='low')
        
        # Apply filter to both channels
        distance_audio[:, 0] = signal.filtfilt(b, a, distance_audio[:, 0])
        distance_audio[:, 1] = signal.filtfilt(b, a, distance_audio[:, 1])
    
    return distance_audio

def apply_looming_effect(audio, distance, max_distance):
    """
    Apply looming effect to enhance the perception of an approaching sound
    
    Looming involves spectral and amplitude changes that create a sense
    of urgency as objects approach
    """
    if not ENABLE_LOOMING:
        return audio
    
    # Normalize distance to 0-1 range (1 = furthest, 0 = closest)
    norm_distance = distance / max_distance
    
    # Create enhanced audio
    looming_audio = audio.copy()
    
    # 1. Enhanced amplitude changes (nonlinear)
    # Approach sounds are perceived as louder than expected from inverse square law
    looming_factor = 1.0 + (1.0 - norm_distance) ** 2
    looming_audio *= looming_factor
    
    # 2. Spectral emphasis changes with distance
    # As objects approach, mid-high frequencies become more prominent
    if norm_distance < 0.8:  # Only apply for closer sounds
        # Boost mid and high frequencies as object gets closer
        # Use a custom method instead of highshelf filter
        boost_amount = (1.0 - norm_distance) * 6.0  # dB boost (max 6dB)
        shelf_freq = 1000  # Hz
        
        # Simple frequency boost using FFT
        for ch in range(looming_audio.shape[1]):
            # Get channel data
            channel_data = looming_audio[:, ch]
            
            # Apply FFT
            fft_data = np.fft.rfft(channel_data)
            freqs = np.fft.rfftfreq(len(channel_data), 1.0/SAMPLE_RATE)
            
            # Create frequency-dependent gain
            gain_factor = np.ones(len(freqs))
            
            # Apply boost to frequencies above shelf_freq
            above_shelf = freqs >= shelf_freq
            # Convert dB to amplitude ratio (dB = 20 * log10(ratio))
            boost_ratio = 10 ** (boost_amount / 20)
            
            # Create a gradual transition
            transition_width = 200  # Hz
            for i, freq in enumerate(freqs):
                if freq >= shelf_freq:
                    # Linear transition from 1.0 to boost_ratio over transition_width
                    if freq < shelf_freq + transition_width:
                        transition_factor = (freq - shelf_freq) / transition_width
                        gain_factor[i] = 1.0 + (boost_ratio - 1.0) * transition_factor
                    else:
                        gain_factor[i] = boost_ratio
            
            # Apply gain
            fft_data *= gain_factor
            
            # Apply inverse FFT
            looming_audio[:, ch] = np.real(np.fft.irfft(fft_data, len(channel_data)))
    
    # 3. Add slight harmonic distortion for very close sounds (creates urgency)
    if norm_distance < 0.3:
        # Add subtle harmonic distortion proportional to proximity
        distortion_amount = (0.3 - norm_distance) * 0.1
        
        # Simple waveshaping distortion
        looming_audio = np.tanh(looming_audio * (1.0 + distortion_amount)) / (1.0 + distortion_amount)
    
    # Normalize to avoid clipping
    max_val = np.max(np.abs(looming_audio))
    if max_val > 1.0:
        looming_audio /= max_val
    
    return looming_audio

def apply_doppler_effect(source_signal, distances, fs=SAMPLE_RATE):
    """
    Apply Doppler effect to the audio based on the source's velocity
    
    The Doppler effect causes frequency shifts based on relative velocity:
    - Approaching sound sources have higher pitch
    - Receding sound sources have lower pitch
    """
    if not ENABLE_DOPPLER or len(distances) < 2:
        return source_signal
    
    try:
        # Calculate time step
        time_step = DURATION / len(distances)
        
        # Calculate velocities (m/s) - negative velocity means approaching
        velocities = np.zeros(len(distances) - 1)
        for i in range(len(distances) - 1):
            velocities[i] = (distances[i+1] - distances[i]) / time_step
        
        # Add a final velocity (repeat the last one)
        velocities = np.append(velocities, velocities[-1])
        
        # Calculate the Doppler shift factor: f' = f * (v_sound / (v_sound + v_source))
        # Note: approaching source has negative velocity, so denominator decreases
        doppler_factors = SPEED_OF_SOUND / (SPEED_OF_SOUND + velocities)
        
        # Limit the maximum frequency shift to avoid extreme effects
        doppler_factors = np.clip(doppler_factors, 0.5, 2.0)
        
        # Get length of source signal
        n_samples = len(source_signal)
        
        # Initialize output signal
        output_signal = np.zeros_like(source_signal)
        
        # Calculate resampling indices for each sample
        # This maps where each output sample should come from in the input
        resample_indices = np.zeros(n_samples)
        current_index = 0.0
        
        segment_length = n_samples // len(doppler_factors)
        
        for i in range(n_samples):
            # Determine which segment we're in
            segment_idx = min(i // segment_length, len(doppler_factors) - 1)
            
            # Get the doppler factor for this segment
            doppler_factor = doppler_factors[segment_idx]
            
            # Update the current index
            resample_indices[i] = current_index
            current_index += doppler_factor
        
        # Clip indices to valid range
        max_idx = n_samples - 1
        resample_indices = np.clip(resample_indices, 0, max_idx)
        
        # Create interpolated output using linear interpolation
        for ch in range(source_signal.shape[1]):
            output_signal[:, ch] = np.interp(
                resample_indices, 
                np.arange(n_samples), 
                source_signal[:, ch]
            )
        
        return output_signal
        
    except Exception as e:
        print(f"Error applying Doppler effect: {e}")
        import traceback
        traceback.print_exc()
        return source_signal

def create_approaching_movement(hrtf, source_signal, room=None, verbose=True):
    """
    Create audio where the sound stays at constant azimuth (left)
    but approaches the listener from far to near
    """
    # Calculate total samples
    total_samples = len(source_signal)
    segment_samples = total_samples // N_SEGMENTS
    
    # Create distance trajectory (from far to near)
    distances = np.linspace(INITIAL_DISTANCE, FINAL_DISTANCE, N_SEGMENTS)
    
    # Create output buffer
    output_signal = np.zeros((total_samples, 2))
    
    # We'll use the same HRTF for all segments (constant angle)
    # but will apply distance-based effects to simulate the approach
    
    # Find the HRTF for our target angle
    nearest_idx, nearest_pos = find_nearest_hrtf(hrtf, CONSTANT_AZIMUTH, CONSTANT_ELEVATION)
    hrtf_left, hrtf_right = extract_hrtf_data(hrtf, nearest_idx)
    
    if verbose:
        if nearest_pos is not None:
            if hrtf.SourcePosition_Type == 'spherical' and np.max(np.abs(nearest_pos[0])) > 6.28:
                pos_str = f"({nearest_pos[0]:.1f}°, {nearest_pos[1]:.1f}°)"
            elif hrtf.SourcePosition_Type == 'spherical':
                pos_str = f"({np.rad2deg(nearest_pos[0]):.1f}°, {np.rad2deg(nearest_pos[1]):.1f}°)"
            else:
                pos_str = f"[{nearest_pos[0]:.2f}, {nearest_pos[1]:.2f}, {nearest_pos[2]:.2f}]"
        else:
            pos_str = "unknown"
            
        print(f"Using HRTF measurement at position {pos_str} for all segments")
    
    # Apply the HRTF to the entire source signal
    print("Applying HRTF filter...")
    full_left = signal.fftconvolve(source_signal[:, 0], hrtf_left, mode='same')
    full_right = signal.fftconvolve(source_signal[:, 0], hrtf_right, mode='same')
    binaural_base = np.column_stack((full_left, full_right))
    
    # Apply Doppler effect to the entire signal before segmenting
    if ENABLE_DOPPLER:
        print("Applying Doppler effect...")
        binaural_base = apply_doppler_effect(binaural_base, distances)
    
    # Now process each segment with the appropriate distance model
    if verbose:
        print(f"Processing {N_SEGMENTS} segments with distance effects...")
    
    for i in range(N_SEGMENTS):
        # Calculate segment boundaries
        start_idx = i * segment_samples
        end_idx = min((i + 1) * segment_samples, total_samples)
        
        # Get the binaural segment
        segment = binaural_base[start_idx:end_idx]
        
        # Get distance for this segment
        distance = distances[i]
        
        # 1. Apply basic distance model (amplitude and filtering)
        processed_segment = apply_basic_distance_model(segment, distance)
        
        # 2. Apply looming effect if enabled
        if ENABLE_LOOMING:
            processed_segment = apply_looming_effect(processed_segment, distance, INITIAL_DISTANCE)
        
        # 3. Apply room acoustics if enabled
        if ENABLE_ROOM_ACOUSTICS and room is not None:
            # Convert spherical coordinates to Cartesian for room model
            # For -90° azimuth (left), this is (-distance, 0)
            source_position = [distance * np.cos(np.deg2rad(CONSTANT_AZIMUTH + 90)), 
                             distance * np.sin(np.deg2rad(CONSTANT_AZIMUTH + 90))]
            
            # Process with room acoustics
            # Note: We'll skip this for now and apply room acoustics only to final output
            # This is because applying room acoustics to each segment is computationally intensive
            pass
        
        # Add to output buffer
        output_signal[start_idx:end_idx] = processed_segment
        
        # Print progress
        if verbose and (i % 5 == 0 or i == N_SEGMENTS - 1):
            print(f"Processed segment {i+1}/{N_SEGMENTS} at distance {distance:.2f}m")
    
    # Apply a gentle smoothing between segments to avoid clicks
    if N_SEGMENTS > 1:
        print("Applying final smoothing...")
        
        # Use a short crossfade at segment boundaries
        crossfade_len = min(segment_samples // 10, 200)
        
        for i in range(1, N_SEGMENTS):
            boundary_idx = i * segment_samples
            
            # Make sure we have enough samples for crossfade
            if boundary_idx - crossfade_len // 2 >= 0 and boundary_idx + crossfade_len // 2 < total_samples:
                # Get the audio around the boundary
                start_idx = boundary_idx - crossfade_len // 2
                end_idx = boundary_idx + crossfade_len // 2
                
                # Ensure we have an even length for the window
                window_len = end_idx - start_idx
                
                # Get the boundary audio segment
                boundary_audio = output_signal[start_idx:end_idx]
                
                # Create a window of exactly the right length
                window = np.hanning(window_len)
                
                # Apply window to both channels
                boundary_audio *= window[:, np.newaxis]
                
                # Put back into the output
                output_signal[start_idx:end_idx] = boundary_audio
    
    # Apply room acoustics to the final output if enabled
    if ENABLE_ROOM_ACOUSTICS and room is not None and PYROOMACOUSTICS_AVAILABLE:
        print("Applying room acoustics to final output...")
        # For demo purposes, use the middle distance position
        mid_distance = (INITIAL_DISTANCE + FINAL_DISTANCE) / 2
        source_position = [mid_distance * np.cos(np.deg2rad(CONSTANT_AZIMUTH + 90)), 
                         mid_distance * np.sin(np.deg2rad(CONSTANT_AZIMUTH + 90))]
        
        # Apply room acoustics
        try:
            output_signal = apply_room_acoustics(output_signal, source_position, room)
            print("Room acoustics applied successfully")
        except Exception as e:
            print(f"Error applying room acoustics: {e}")
    
    # Normalize output
    max_val = np.max(np.abs(output_signal))
    if max_val > 0:
        output_signal = output_signal / max_val * 0.9
    
    return output_signal, distances

def plot_distance_trajectory(distances, output_path):
    """Plot the trajectory of the approaching sound (distance over time)"""
    # Create figure
    plt.figure(figsize=(10, 6))
    
    # Define time axis (in seconds)
    time_axis = np.linspace(0, DURATION, len(distances))
    
    # Plot distance vs time
    plt.plot(time_axis, distances, 'b-', linewidth=3)
    plt.scatter(time_axis, distances, c=distances, cmap='viridis', s=80, alpha=0.7)
    
    # Mark start and end points
    plt.scatter(time_axis[0], distances[0], color='green', s=120, marker='o', label="Start (Far)")
    plt.scatter(time_axis[-1], distances[-1], color='red', s=120, marker='o', label="End (Near)")
    
    # Add grid and labels
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.title(f"Sound Source Approaching from the Left ({CONSTANT_AZIMUTH}° azimuth)", fontsize=14)
    plt.xlabel("Time (seconds)", fontsize=12)
    plt.ylabel("Distance from Listener (meters)", fontsize=12)
    plt.legend(fontsize=10)
    
    # Add annotations
    plt.annotate("Far", (time_axis[0], distances[0]), 
                xytext=(time_axis[0]-0.2, distances[0]+0.2),
                fontsize=12, fontweight='bold')
    plt.annotate("Near", (time_axis[-1], distances[-1]), 
                xytext=(time_axis[-1]-0.2, distances[-1]+0.2),
                fontsize=12, fontweight='bold')
    
    # Add top-down view of the scene
    ax_inset = plt.axes([0.65, 0.15, 0.3, 0.3])  # [left, bottom, width, height]
    
    # Plot listener (center) and sound positions
    ax_inset.scatter([0], [0], color='black', s=200, marker='o', label="Listener")
    
    # Calculate positions along azimuth at different distances
    # For -90° azimuth, x=0, y=-distance
    # We rotate by 90° to convert from azimuth to Cartesian
    x_positions = distances * np.cos(np.deg2rad(CONSTANT_AZIMUTH + 90))
    y_positions = distances * np.sin(np.deg2rad(CONSTANT_AZIMUTH + 90))
    
    # Plot trajectory points
    ax_inset.scatter(x_positions, y_positions, c=np.arange(len(distances)), cmap='viridis', s=50, alpha=0.7)
    ax_inset.plot(x_positions, y_positions, 'k-', alpha=0.5)
    
    # Mark start and end
    ax_inset.scatter(x_positions[0], y_positions[0], color='green', s=80, marker='o')
    ax_inset.scatter(x_positions[-1], y_positions[-1], color='red', s=80, marker='o')
    
    # Add an arrow to show direction of movement
    arrow_idx = len(distances) // 2
    ax_inset.arrow(x_positions[arrow_idx], y_positions[arrow_idx], 
                  (x_positions[arrow_idx+1] - x_positions[arrow_idx])*3, 
                  (y_positions[arrow_idx+1] - y_positions[arrow_idx])*3,
                  head_width=0.1, head_length=0.15, fc='red', ec='red')
    
    # Draw room boundary if room acoustics is enabled
    if ENABLE_ROOM_ACOUSTICS:
        circle = plt.Circle((0, 0), ROOM_RADIUS, fill=False, linestyle='--', color='gray', alpha=0.5)
        ax_inset.add_artist(circle)
        ax_inset.text(0, ROOM_RADIUS, "Room", ha='center', va='bottom', 
                     color='gray', fontsize=8, alpha=0.7)
    
    # Set equal aspect ratio for inset
    ax_inset.set_aspect('equal')
    ax_inset.set_title("Top-Down View", fontsize=10)
    ax_inset.grid(True, linestyle='--', alpha=0.3)
    ax_inset.set_xlabel("X (meters)")
    ax_inset.set_ylabel("Y (meters)")
    
    # Set axis limits to show the room
    max_range = max(abs(INITIAL_DISTANCE), abs(FINAL_DISTANCE), ROOM_RADIUS) * 1.2
    ax_inset.set_xlim(-max_range, max_range)
    ax_inset.set_ylim(-max_range, max_range)
    
    # Save figure
    plt.savefig(output_path, dpi=300, bbox_inches='tight')
    plt.close()
    
    print(f"Saved trajectory plot to: {output_path}")

def main():
    """Generate approaching-from-left audio for each SOFA model with enhanced effects"""
    if not SOFA_AVAILABLE:
        print("ERROR: sofar library is required. Please install it with:")
        print("pip install sofar")
        return
    
    # Set up the room model if room acoustics is enabled
    room = None
    if ENABLE_ROOM_ACOUSTICS and PYROOMACOUSTICS_AVAILABLE:
        room = setup_room_model()
    
    # Generate source signal (stereo)
    print("Generating source signal...")
    mono_source = generate_pink_noise(DURATION, SAMPLE_RATE)
    source_signal = np.column_stack((mono_source, mono_source))
    
    # Process each SOFA file
    results = []
    valid_files = []
    
    for sofa_path in sofa_files:
        if not os.path.exists(sofa_path):
            print(f"File not found: {sofa_path}")
            continue
        
        valid_files.append(sofa_path)
        file_name = os.path.basename(sofa_path).replace('.sofa', '')
        print(f"\nProcessing {file_name}...")
        
        try:
            # Load SOFA file
            hrtf, dimensions = load_sofa_file(sofa_path)
            
            if hrtf is None:
                print(f"Failed to load {file_name}, skipping.")
                continue
            
            # Create approaching movement with enhanced effects
            output_signal, distances = create_approaching_movement(hrtf, source_signal, room)
            
            # Save audio file
            timestamp = time.strftime("%Y%m%d_%H%M%S")
            effect_indicators = []
            if ENABLE_ROOM_ACOUSTICS: effect_indicators.append("room")
            if ENABLE_DOPPLER: effect_indicators.append("doppler")
            if ENABLE_LOOMING: effect_indicators.append("looming")
            effects_str = "_".join(effect_indicators) if effect_indicators else "basic"
            
            output_filename = f"approaching_{file_name}_{effects_str}_{timestamp}.wav"
            output_path = os.path.join(output_dir, output_filename)
            sf.write(output_path, output_signal, SAMPLE_RATE)
            print(f"Saved audio to: {output_path}")
            
            # Store result for comparison
            results.append({
                'name': file_name,
                'signal': output_signal,
                'path': output_path
            })
            
        except Exception as e:
            print(f"Error processing {file_name}: {e}")
            import traceback
            traceback.print_exc()
    
    # Create trajectory plot (only need to do this once)
    if len(results) > 0:
        effects_str = ""
        if ENABLE_ROOM_ACOUSTICS: effects_str += "_room"
        if ENABLE_DOPPLER: effects_str += "_doppler"
        if ENABLE_LOOMING: effects_str += "_looming"
        
        plot_path = os.path.join(output_dir, 
                               f"approaching_trajectory{effects_str}_{time.strftime('%Y%m%d_%H%M%S')}.png")
        plot_distance_trajectory(distances, plot_path)
    
    print("\nApproaching movement processing complete!")
    print(f"Total SOFA models processed: {len(results)} of {len(valid_files)}")
    print(f"All files saved to: {output_dir}")
    
    # Summarize settings used
    print("\nSettings used:")
    print(f"  Initial distance: {INITIAL_DISTANCE}m")
    print(f"  Final distance: {FINAL_DISTANCE}m")
    print(f"  Motion duration: {DURATION}s")
    print(f"  Azimuth: {CONSTANT_AZIMUTH}°")
    print(f"  Elevation: {CONSTANT_ELEVATION}°")
    print(f"  Room acoustics: {'Enabled' if ENABLE_ROOM_ACOUSTICS else 'Disabled'}")
    if ENABLE_ROOM_ACOUSTICS:
        print(f"    Room radius: {ROOM_RADIUS}m")
        print(f"    RT60: {RT60}s")
    print(f"  Doppler effect: {'Enabled' if ENABLE_DOPPLER else 'Disabled'}")
    print(f"  Looming effect: {'Enabled' if ENABLE_LOOMING else 'Disabled'}")
    
    # List all generated files
    print("\nGenerated files:")
    for result in results:
        print(f"- {os.path.basename(result['path'])}")

if __name__ == "__main__":
    main()
```

```python
"""
Enhanced SOFA HRTF Distance-Based Movement

This script creates binaural audio files where a sound source approaches the listener
at ear level (0° elevation), maintaining the same angle but decreasing in distance,
for various HRTF models. This version creates variants with different azimuth angles.

The script includes room acoustics, looming effect, and Doppler effect to create
a more realistic approaching sound with smooth transitions.
"""
import numpy as np
import scipy.signal as signal
import soundfile as sf
import matplotlib.pyplot as plt
import os
import time
import datetime

# First, ensure required libraries are installed
try:
    import sofar
    SOFA_AVAILABLE = True
except ImportError:
    print("Warning: sofar library not found. Please install with: pip install sofar")
    SOFA_AVAILABLE = False

try:
    import pyroomacoustics as pra
    PYROOMACOUSTICS_AVAILABLE = True
except ImportError:
    print("Warning: pyroomacoustics library not found. Please install with: pip install pyroomacoustics")
    PYROOMACOUSTICS_AVAILABLE = False

#############################################################
# CONTROL PARAMETERS - ADJUST THESE TO CHANGE THE SIMULATION
#############################################################

# Spatial parameters
INITIAL_DISTANCE = 2.0    # Starting distance in meters
FINAL_DISTANCE = 0.3      # Final distance in meters (closest approach)
AZIMUTH_ANGLES = [90, 0, -90]  # Angles to generate (right, front, left)
CONSTANT_ELEVATION = 0    # Fixed elevation (ear level)

# Time parameters
DURATION = 3.0            # Duration of movement in seconds (longer = slower movement)
SAMPLE_RATE = 44100       # Audio sample rate in Hz
N_SEGMENTS = 60           # Number of segments for processing (higher = smoother movement)
FADE_OUT_DURATION = 0.2   # Duration of the fade-out at the end (seconds)

# Room acoustics parameters
ROOM_RADIUS = 2.0         # Radius of the spherical room in meters
WALL_ABSORPTION = 0.3     # Wall absorption coefficient (0-1, higher = more absorption)
RT60 = 0.4                # Desired RT60 reverberation time (seconds)

# Effect toggles
ENABLE_ROOM_ACOUSTICS = True   # Whether to simulate room acoustics
ENABLE_DOPPLER = True          # Whether to simulate Doppler effect 
ENABLE_LOOMING = True          # Whether to enhance approaching sound with looming effect

# Filter for only CTF files
sofa_files = [
    r"C:\Users\cogpsy-vrlab\Downloads\HRIRs_neutral_head_orientation_v4\SOFA\FABIAN_CTF_measured.sofa",
    r"C:\Users\cogpsy-vrlab\Downloads\HRIRs_neutral_head_orientation_v4\SOFA\FABIAN_CTF_measured_smoothed.sofa"
]

# Create output directory with timestamp
timestamp = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
output_dir = os.path.join(r"C:\Users\cogpsy-vrlab\Documents\PPS_module\BreathingPilot\LoomingStimuli", f"looming_stimuli_{timestamp}")
os.makedirs(output_dir, exist_ok=True)

# Calculate speed of sound (m/s)
SPEED_OF_SOUND = 343.0

def generate_pink_noise(duration, fs):
    """Generate pink noise using FFT-based method"""
    n_samples = int(duration * fs)
    
    # Generate white noise
    np.random.seed(42)  # For reproducibility
    white_noise = np.random.randn(n_samples)
    
    # Get the FFT of the white noise
    X = np.fft.rfft(white_noise)
    
    # Create pink filter (1/f)
    n_bins = len(X)
    f = np.linspace(1, n_bins, n_bins)
    f[0] = 1  # Avoid division by zero
    
    # Apply 1/sqrt(f) filter for power spectrum ~ 1/f
    X = X / np.sqrt(f)
    
    # Convert back to time domain
    pink_noise = np.real(np.fft.irfft(X, n_samples))
    
    # Normalize
    pink_noise = pink_noise / np.max(np.abs(pink_noise)) * 0.8
    
    # Apply fade in/out to avoid clicks
    fade_samples = int(0.05 * fs)  # 50ms fade
    fade_in = np.linspace(0, 1, fade_samples)
    fade_out = np.linspace(1, 0, fade_samples)
    
    pink_noise[:fade_samples] *= fade_in
    pink_noise[-fade_samples:] *= fade_out
    
    return pink_noise

def load_sofa_file(sofa_path):
    """Load HRTF data from SOFA file"""
    try:
        print(f"Loading SOFA file: {os.path.basename(sofa_path)}")
        
        # Load the SOFA file
        hrtf = sofar.read_sofa(sofa_path)
        
        # Display basic SOFA info
        print(f"SOFA convention: {hrtf.GLOBAL_SOFAConventions} v{hrtf.GLOBAL_SOFAConventionsVersion}")
        
        # Get dimensions
        dimensions = {}
        for dim_name in ['M', 'R', 'N']:
            try:
                dimensions[dim_name] = hrtf.get_dimension(dim_name)
                print(f"{dim_name} = {dimensions[dim_name]}")
            except:
                dimensions[dim_name] = None
                print(f"{dim_name} = Unknown")
        
        return hrtf, dimensions
        
    except Exception as e:
        print(f"Error loading SOFA file: {e}")
        import traceback
        traceback.print_exc()
        return None, None

def find_nearest_hrtf(hrtf, target_azimuth, target_elevation=0, target_distance=1.0):
    """Find the measurement closest to the target azimuth, elevation, and distance"""
    try:
        source_positions = np.array(hrtf.SourcePosition)
        
        # Handle different position array shapes
        if source_positions.shape[0] == 3 and len(source_positions.shape) == 2:
            source_positions = source_positions.T
        
        # Check if positions are in degrees or radians
        position_type = hrtf.SourcePosition_Type
        max_azimuth = np.max(np.abs(source_positions[:, 0]))
        positions_in_degrees = max_azimuth > 6.28
        
        if positions_in_degrees:
            # Both source positions and target are in degrees
            az_compare = target_azimuth
            el_compare = target_elevation
        else:
            # Source positions in radians, convert target to radians
            az_compare = np.deg2rad(target_azimuth)
            el_compare = np.deg2rad(target_elevation)
        
        # Handle negative azimuths if data is in 0-360 format
        if positions_in_degrees and target_azimuth < 0:
            # Check if we have negative azimuths in the data
            has_negative_az = np.any(source_positions[:, 0] < 0)
            if not has_negative_az:
                # Convert target from -180 to 180 format to 0 to 360 format
                az_compare = (target_azimuth + 360) % 360
        
        # Create distance metric
        # For spherical coordinates: focus mostly on azimuth, then elevation
        # Ignore distance since most HRTF databases only have one distance
        if position_type == 'spherical':
            dists = np.sqrt(
                2 * (source_positions[:, 0] - az_compare)**2 + 
                (source_positions[:, 1] - el_compare)**2
            )
        else:
            # For Cartesian: calculate actual 3D distance
            # Create a target position in Cartesian
            if position_type == 'cartesian':
                # Convert spherical to Cartesian
                target_x = target_distance * np.cos(np.deg2rad(target_elevation)) * np.cos(np.deg2rad(target_azimuth))
                target_y = target_distance * np.cos(np.deg2rad(target_elevation)) * np.sin(np.deg2rad(target_azimuth))
                target_z = target_distance * np.sin(np.deg2rad(target_elevation))
                target_pos = np.array([target_x, target_y, target_z])
                
                # Calculate Euclidean distance to each position
                dists = np.sqrt(np.sum((source_positions - target_pos)**2, axis=1))
            else:
                # Unknown position type, use simple difference
                dists = np.abs(source_positions[:, 0] - az_compare)
                
        # Find nearest position
        nearest_idx = np.argmin(dists)
        nearest_pos = source_positions[nearest_idx]
        
        return nearest_idx, nearest_pos
        
    except Exception as e:
        print(f"Error finding nearest HRTF: {e}")
        import traceback
        traceback.print_exc()
        return 0, None

def extract_hrtf_data(hrtf, idx):
    """Extract HRTF impulse responses for a given index"""
    hrtf_data = hrtf.Data_IR
    
    # Check dimensions and extract
    if hrtf_data.shape[0] == 2 and hrtf_data.shape[1] > 2:
        # Shape is (R, M, N)
        hrtf_left = hrtf_data[0, idx, :]
        hrtf_right = hrtf_data[1, idx, :]
    else:
        # Shape is (M, R, N)
        try:
            hrtf_left = hrtf_data[idx, 0, :]
            hrtf_right = hrtf_data[idx, 1, :]
        except:
            # Try other dimension orderings if the above fails
            try:
                if len(hrtf_data.shape) == 4:  # If 4D array
                    hrtf_left = hrtf_data[idx, 0, 0, :]
                    hrtf_right = hrtf_data[idx, 1, 0, :]
                else:
                    raise ValueError(f"Unexpected HRTF data shape: {hrtf_data.shape}")
            except Exception as e:
                print(f"Error extracting HRTF data: {e}")
                # Return empty IRs as fallback
                dummy_ir = np.zeros(256)
                dummy_ir[0] = 1.0
                return dummy_ir, dummy_ir
    
    return hrtf_left, hrtf_right

def setup_room_model():
    """Create a spherical room model using pyroomacoustics"""
    if not PYROOMACOUSTICS_AVAILABLE or not ENABLE_ROOM_ACOUSTICS:
        print("Room acoustics disabled or pyroomacoustics not available")
        return None
    
    try:
        # Calculate actual RT60 based on room parameters
        # Using Sabine's formula: RT60 = 0.161 * V / (A * S)
        # where V is volume, A is absorption coefficient, S is surface area
        room_volume = (4/3) * np.pi * ROOM_RADIUS**3
        room_surface = 4 * np.pi * ROOM_RADIUS**2
        
        # Adjust absorption to match desired RT60 if possible
        sabine_absorption = 0.161 * room_volume / (RT60 * room_surface)
        
        # Clamp absorption between 0.01 and 0.99
        absorption = min(0.99, max(0.01, sabine_absorption))
        
        print(f"Creating spherical room model with radius {ROOM_RADIUS}m")
        print(f"Calculated absorption coefficient: {absorption:.2f} for RT60 of {RT60}s")
        
        # Create a circular room approximation (2D projection of sphere)
        num_walls = 16
        angles = np.linspace(0, 2*np.pi, num_walls, endpoint=False)
        corners = np.array([[ROOM_RADIUS * np.cos(ang), ROOM_RADIUS * np.sin(ang)] 
                          for ang in angles]).T
        
        # Create room with the specified absorption
        materials = pra.Material(energy_absorption=absorption)
        room = pra.Room.from_corners(corners, fs=SAMPLE_RATE, materials=materials, max_order=3)
        
        # Create stereo microphone array (listener's ears)
        ear_spacing = 0.18  # typical head size
        mic_positions = np.array([
            [-ear_spacing/2, 0],  # Left ear
            [ear_spacing/2, 0]    # Right ear
        ]).T
        
        mic_array = pra.MicrophoneArray(mic_positions, room.fs)
        room.add_microphone_array(mic_array)
        
        return room
    
    except Exception as e:
        print(f"Error setting up room model: {e}")
        import traceback
        traceback.print_exc()
        return None

def apply_room_acoustics(audio, source_position, room):
    """Apply room acoustics to the audio using the pyroomacoustics model"""
    if room is None:
        return audio
    
    try:
        # Reset any existing sources
        room.sources = []
        
        # Add new source with the given position and mono audio
        # For simplicity, use the average of stereo channels
        if audio.ndim == 2:
            mono_signal = np.mean(audio, axis=1)
        else:
            mono_signal = audio
        
        # Add the source to the room
        room.add_source(source_position, signal=mono_signal)
        
        # Compute the room impulse responses
        room.compute_rir()
        
        # Simulate the room acoustics
        room_output = room.simulate()
        
        # Convert to stereo output
        output_signal = np.zeros((len(room_output), 2))
        output_signal[:, 0] = room_output[:, 0]  # Left ear
        output_signal[:, 1] = room_output[:, 1]  # Right ear
        
        # Normalize output
        max_val = np.max(np.abs(output_signal))
        if max_val > 0:
            output_signal = output_signal / max_val * 0.9
            
        return output_signal
        
    except Exception as e:
        print(f"Error applying room acoustics: {e}")
        import traceback
        traceback.print_exc()
        return audio

def apply_basic_distance_model(audio, distance, reference_distance=1.0, final_distance=FINAL_DISTANCE):
    """
    Apply a basic distance model to simulate source distance with improved transition at close distances
    
    This uses a modified inverse square law for amplitude and
    adds a low-pass filter effect for further distances
    """
    # Apply a modified inverse square law for amplitude that limits extreme amplification at close distances
    distance_ratio = max(0.1, min(distance / reference_distance, 10.0))
    
    # Modified amplitude scaling that softens the transition as we get very close
    # Use a nonlinear curve that asymptotically approaches a maximum value
    if distance <= final_distance * 1.5:
        # For distances very close to the final distance, use a gentler curve
        min_ratio = final_distance / reference_distance
        normalized_ratio = (distance_ratio - min_ratio) / (1.0 - min_ratio)
        # Use a sigmoid-like function to smooth the transition
        transition_factor = 0.5 * (1 - np.cos(normalized_ratio * np.pi))
        amplitude_scale = 1.0 / (distance_ratio ** (1.5 * transition_factor))
    else:
        # Standard inverse square law for further distances
        amplitude_scale = 1.0 / (distance_ratio ** 2)
    
    # Normalize scale to avoid extreme amplification
    amplitude_scale = min(amplitude_scale, 2.5)
    distance_audio = audio * amplitude_scale
    
    # Apply distance-based filtering (closer = brighter, further = more muffled)
    if distance > reference_distance:
        # Create a lowpass filter for distant sounds
        cutoff_freq = max(1000, min(20000, 20000 / distance_ratio))
        cutoff_norm = cutoff_freq / (SAMPLE_RATE/2)  # Normalize to Nyquist
        b, a = signal.butter(2, cutoff_norm, btype='low')
        
        # Apply filter to both channels
        distance_audio[:, 0] = signal.filtfilt(b, a, distance_audio[:, 0])
        distance_audio[:, 1] = signal.filtfilt(b, a, distance_audio[:, 1])
    
    return distance_audio

def apply_looming_effect(audio, distance, max_distance):
    """
    Apply looming effect to enhance the perception of an approaching sound
    
    Looming involves spectral and amplitude changes that create a sense
    of urgency as objects approach
    """
    if not ENABLE_LOOMING:
        return audio
    
    # Normalize distance to 0-1 range (1 = furthest, 0 = closest)
    norm_distance = distance / max_distance
    
    # Create enhanced audio
    looming_audio = audio.copy()
    
    # 1. Enhanced amplitude changes (nonlinear)
    # Approach sounds are perceived as louder than expected from inverse square law
    # Using a gentler curve for close distances to prevent the jump
    if norm_distance < 0.2:
        # Softer enhancement for very close distances
        looming_factor = 1.0 + (1.0 - norm_distance) * 0.5
    else:
        looming_factor = 1.0 + (1.0 - norm_distance) ** 1.5
    
    looming_audio *= looming_factor
    
    # 2. Spectral emphasis changes with distance
    # As objects approach, mid-high frequencies become more prominent
    if norm_distance < 0.8:  # Only apply for closer sounds
        # Boost mid and high frequencies as object gets closer
        # Use a custom method instead of highshelf filter
        boost_amount = (1.0 - norm_distance) * 5.0  # dB boost (max 5dB, reduced from 6dB)
        shelf_freq = 1000  # Hz
        
        # Simple frequency boost using FFT
        for ch in range(looming_audio.shape[1]):
            # Get channel data
            channel_data = looming_audio[:, ch]
            
            # Apply FFT
            fft_data = np.fft.rfft(channel_data)
            freqs = np.fft.rfftfreq(len(channel_data), 1.0/SAMPLE_RATE)
            
            # Create frequency-dependent gain
            gain_factor = np.ones(len(freqs))
            
            # Apply boost to frequencies above shelf_freq
            above_shelf = freqs >= shelf_freq
            # Convert dB to amplitude ratio (dB = 20 * log10(ratio))
            boost_ratio = 10 ** (boost_amount / 20)
            
            # Create a gradual transition
            transition_width = 200  # Hz
            for i, freq in enumerate(freqs):
                if freq >= shelf_freq:
                    # Linear transition from 1.0 to boost_ratio over transition_width
                    if freq < shelf_freq + transition_width:
                        transition_factor = (freq - shelf_freq) / transition_width
                        gain_factor[i] = 1.0 + (boost_ratio - 1.0) * transition_factor
                    else:
                        gain_factor[i] = boost_ratio
            
            # Apply gain
            fft_data *= gain_factor
            
            # Apply inverse FFT
            looming_audio[:, ch] = np.real(np.fft.irfft(fft_data, len(channel_data)))
    
    # 3. Add slight harmonic distortion for very close sounds (creates urgency)
    if norm_distance < 0.3:
        # Reduce distortion for very close distances to avoid the jump at the end
        if norm_distance < 0.1:
            # Gradually reduce distortion as we get extremely close
            distortion_amount = (0.1 - norm_distance) * 0.03  # Much gentler distortion
        else:
            distortion_amount = (0.3 - norm_distance) * 0.07  # Reduced from 0.1
        
        # Simple waveshaping distortion
        looming_audio = np.tanh(looming_audio * (1.0 + distortion_amount)) / (1.0 + distortion_amount)
    
    # Normalize to avoid clipping
    max_val = np.max(np.abs(looming_audio))
    if max_val > 1.0:
        looming_audio /= max_val
    
    return looming_audio

def apply_doppler_effect(source_signal, distances, fs=SAMPLE_RATE):
    """
    Apply Doppler effect to the audio based on the source's velocity
    
    The Doppler effect causes frequency shifts based on relative velocity:
    - Approaching sound sources have higher pitch
    - Receding sound sources have lower pitch
    """
    if not ENABLE_DOPPLER or len(distances) < 2:
        return source_signal
    
    try:
        # Calculate time step
        time_step = DURATION / len(distances)
        
        # Calculate velocities (m/s) - negative velocity means approaching
        velocities = np.zeros(len(distances) - 1)
        for i in range(len(distances) - 1):
            velocities[i] = (distances[i+1] - distances[i]) / time_step
        
        # Add a final velocity (repeat the last one)
        velocities = np.append(velocities, velocities[-1])
        
        # Smooth velocities to avoid abrupt changes, especially at the end
        window_size = 5
        padded_velocities = np.pad(velocities, (window_size//2, window_size//2), mode='edge')
        smoothed_velocities = np.zeros_like(velocities)
        
        for i in range(len(velocities)):
            smoothed_velocities[i] = np.mean(padded_velocities[i:i+window_size])
        
        velocities = smoothed_velocities
        
        # Calculate the Doppler shift factor: f' = f * (v_sound / (v_sound + v_source))
        # Note: approaching source has negative velocity, so denominator decreases
        doppler_factors = SPEED_OF_SOUND / (SPEED_OF_SOUND + velocities)
        
        # Limit the maximum frequency shift to avoid extreme effects
        doppler_factors = np.clip(doppler_factors, 0.5, 2.0)
        
        # Get length of source signal
        n_samples = len(source_signal)
        
        # Initialize output signal
        output_signal = np.zeros_like(source_signal)
        
        # Calculate resampling indices for each sample
        # This maps where each output sample should come from in the input
        resample_indices = np.zeros(n_samples)
        current_index = 0.0
        
        segment_length = n_samples // len(doppler_factors)
        
        for i in range(n_samples):
            # Determine which segment we're in
            segment_idx = min(i // segment_length, len(doppler_factors) - 1)
            
            # Get the doppler factor for this segment
            doppler_factor = doppler_factors[segment_idx]
            
            # Update the current index
            resample_indices[i] = current_index
            current_index += doppler_factor
        
        # Clip indices to valid range
        max_idx = n_samples - 1
        resample_indices = np.clip(resample_indices, 0, max_idx)
        
        # Create interpolated output using linear interpolation
        for ch in range(source_signal.shape[1]):
            output_signal[:, ch] = np.interp(
                resample_indices, 
                np.arange(n_samples), 
                source_signal[:, ch]
            )
        
        return output_signal
        
    except Exception as e:
        print(f"Error applying Doppler effect: {e}")
        import traceback
        traceback.print_exc()
        return source_signal

def create_nonlinear_distance_trajectory():
    """
    Create a nonlinear distance trajectory that decelerates as it approaches the listener
    This provides a smoother approach at the end, preventing abrupt changes
    """
    # Create a basic linear trajectory
    linear_distances = np.linspace(INITIAL_DISTANCE, FINAL_DISTANCE, N_SEGMENTS)
    
    # Apply an easing function to create a more natural deceleration
    # Using a quadratic ease-out function
    normalized_indices = np.linspace(0, 1, N_SEGMENTS)
    easing_factors = 1 - (1 - normalized_indices) ** 2
    
    # Apply the easing to the linear distances
    distances = INITIAL_DISTANCE - easing_factors * (INITIAL_DISTANCE - FINAL_DISTANCE)
    
    return distances

def create_approaching_movement(hrtf, source_signal, azimuth, room=None, verbose=True):
    """
    Create audio where the sound stays at constant azimuth
    but approaches the listener from far to near
    """
    # Calculate total samples
    total_samples = len(source_signal)
    segment_samples = total_samples // N_SEGMENTS
    
    # Create nonlinear distance trajectory (decelerates as it approaches)
    distances = create_nonlinear_distance_trajectory()
    
    # Create output buffer
    output_signal = np.zeros((total_samples, 2))
    
    # We'll use the same HRTF for all segments (constant angle)
    # but will apply distance-based effects to simulate the approach
    
    # Find the HRTF for our target angle
    nearest_idx, nearest_pos = find_nearest_hrtf(hrtf, azimuth, CONSTANT_ELEVATION)
    hrtf_left, hrtf_right = extract_hrtf_data(hrtf, nearest_idx)
    
    if verbose:
        if nearest_pos is not None:
            if hrtf.SourcePosition_Type == 'spherical' and np.max(np.abs(nearest_pos[0])) > 6.28:
                pos_str = f"({nearest_pos[0]:.1f}°, {nearest_pos[1]:.1f}°)"
            elif hrtf.SourcePosition_Type == 'spherical':
                pos_str = f"({np.rad2deg(nearest_pos[0]):.1f}°, {np.rad2deg(nearest_pos[1]):.1f}°)"
            else:
                pos_str = f"[{nearest_pos[0]:.2f}, {nearest_pos[1]:.2f}, {nearest_pos[2]:.2f}]"
        else:
            pos_str = "unknown"
            
        print(f"Using HRTF measurement at position {pos_str} for all segments")
    
    # Apply the HRTF to the entire source signal
    print("Applying HRTF filter...")
    full_left = signal.fftconvolve(source_signal[:, 0], hrtf_left, mode='same')
    full_right = signal.fftconvolve(source_signal[:, 0], hrtf_right, mode='same')
    binaural_base = np.column_stack((full_left, full_right))
    
    # Apply Doppler effect to the entire signal before segmenting
    if ENABLE_DOPPLER:
        print("Applying Doppler effect...")
        binaural_base = apply_doppler_effect(binaural_base, distances)
    
    # Now process each segment with the appropriate distance model
    if verbose:
        print(f"Processing {N_SEGMENTS} segments with distance effects...")
    
    for i in range(N_SEGMENTS):
        # Calculate segment boundaries
        start_idx = i * segment_samples
        end_idx = min((i + 1) * segment_samples, total_samples)
        
        # Get the binaural segment
        segment = binaural_base[start_idx:end_idx]
        
        # Get distance for this segment
        distance = distances[i]
        
        # 1. Apply basic distance model (amplitude and filtering)
        processed_segment = apply_basic_distance_model(segment, distance, final_distance=FINAL_DISTANCE)
        
        # 2. Apply looming effect if enabled
        if ENABLE_LOOMING:
            processed_segment = apply_looming_effect(processed_segment, distance, INITIAL_DISTANCE)
        
        # 3. Apply room acoustics if enabled
        if ENABLE_ROOM_ACOUSTICS and room is not None:
            # Convert spherical coordinates to Cartesian for room model
            source_position = [distance * np.cos(np.deg2rad(azimuth + 90)), 
                             distance * np.sin(np.deg2rad(azimuth + 90))]
            
            # Process with room acoustics
            # Note: We'll skip this for now and apply room acoustics only to final output
            # This is because applying room acoustics to each segment is computationally intensive
            pass
        
        # Add to output buffer
        output_signal[start_idx:end_idx] = processed_segment
        
        # Print progress
        if verbose and (i % 10 == 0 or i == N_SEGMENTS - 1):
            print(f"Processed segment {i+1}/{N_SEGMENTS} at distance {distance:.2f}m")
    
    # Apply a gentle smoothing between segments to avoid clicks
    if N_SEGMENTS > 1:
        print("Applying smoothing between segments...")
        
        # Use a short crossfade at segment boundaries
        crossfade_len = min(segment_samples // 10, 200)
        
        for i in range(1, N_SEGMENTS):
            boundary_idx = i * segment_samples
            
            # Make sure we have enough samples for crossfade
            if boundary_idx - crossfade_len // 2 >= 0 and boundary_idx + crossfade_len // 2 < total_samples:
                # Get the audio around the boundary
                start_idx = boundary_idx - crossfade_len // 2
                end_idx = boundary_idx + crossfade_len // 2
                
                # Ensure we have an even length for the window
                window_len = end_idx - start_idx
                
                # Get the boundary audio segment
                boundary_audio = output_signal[start_idx:end_idx]
                
                # Create a window of exactly the right length
                window = np.hanning(window_len)
                
                # Apply window to both channels
                boundary_audio *= window[:, np.newaxis]
                
                # Put back into the output
                output_signal[start_idx:end_idx] = boundary_audio
    
    # Apply room acoustics to the final output if enabled
    if ENABLE_ROOM_ACOUSTICS and room is not None and PYROOMACOUSTICS_AVAILABLE:
        print("Applying room acoustics to final output...")
        # For demo purposes, use the middle distance position
        mid_distance = (INITIAL_DISTANCE + FINAL_DISTANCE) / 2
        source_position = [mid_distance * np.cos(np.deg2rad(azimuth + 90)), 
                         mid_distance * np.sin(np.deg2rad(azimuth + 90))]
        
        # Apply room acoustics
        try:
            output_signal = apply_room_acoustics(output_signal, source_position, room)
            print("Room acoustics applied successfully")
        except Exception as e:
            print(f"Error applying room acoustics: {e}")
    
    # Apply a final fade-out at the end to ensure smooth termination
    fade_samples = int(FADE_OUT_DURATION * SAMPLE_RATE)
    fade_start = total_samples - fade_samples
    
    # Create fade-out window
    fade_window = np.ones(total_samples)
    fade_window[fade_start:] = np.linspace(1.0, 0.0, fade_samples)
    
    # Apply fade-out to both channels
    output_signal[:, 0] *= fade_window
    output_signal[:, 1] *= fade_window
    
    # Normalize output
    max_val = np.max(np.abs(output_signal))
    if max_val > 0:
        output_signal = output_signal / max_val * 0.9
    
    return output_signal, distances

def plot_distance_trajectory(distances, azimuth, output_path):
    """Plot the trajectory of the approaching sound (distance over time)"""
    # Create figure
    plt.figure(figsize=(10, 6))
    
    # Define time axis (in seconds)
    time_axis = np.linspace(0, DURATION, len(distances))
    
    # Plot distance vs time
    plt.plot(time_axis, distances, 'b-', linewidth=3)
    plt.scatter(time_axis, distances, c=distances, cmap='viridis', s=80, alpha=0.7)
    
    # Mark start and end points
    plt.scatter(time_axis[0], distances[0], color='green', s=120, marker='o', label="Start (Far)")
    plt.scatter(time_axis[-1], distances[-1], color='red', s=120, marker='o', label="End (Near)")
    
    # Add grid and labels
    plt.grid(True, linestyle='--', alpha=0.5)
    
    # Create azimuth direction string
    if azimuth == 0:
        direction = "Front"
    elif azimuth == 90:
        direction = "Right"
    elif azimuth == -90:
        direction = "Left"
    else:
        direction = f"{azimuth}°"
    
    plt.title(f"Sound Source Approaching from the {direction} ({azimuth}° azimuth)", fontsize=14)
    plt.xlabel("Time (seconds)", fontsize=12)
    plt.ylabel("Distance from Listener (meters)", fontsize=12)
    plt.legend(fontsize=10)
    
    # Add annotations
    plt.annotate("Far", (time_axis[0], distances[0]), 
                xytext=(time_axis[0]-0.2, distances[0]+0.2),
                fontsize=12, fontweight='bold')
    plt.annotate("Near", (time_axis[-1], distances[-1]), 
                xytext=(time_axis[-1]-0.2, distances[-1]+0.2),
                fontsize=12, fontweight='bold')
    
    # Add top-down view of the scene
    ax_inset = plt.axes([0.65, 0.15, 0.3, 0.3])  # [left, bottom, width, height]
    
    # Plot listener (center) and sound positions
    ax_inset.scatter([0], [0], color='black', s=200, marker='o', label="Listener")
    
    # Calculate positions along azimuth at different distances
    # We rotate by 90° to convert from azimuth to Cartesian
    x_positions = distances * np.cos(np.deg2rad(azimuth + 90))
    y_positions = distances * np.sin(np.deg2rad(azimuth + 90))
    
    # Plot trajectory points
    ax_inset.scatter(x_positions, y_positions, c=np.arange(len(distances)), cmap='viridis', s=50, alpha=0.7)
    ax_inset.plot(x_positions, y_positions, 'k-', alpha=0.5)
    
    # Mark start and end
    ax_inset.scatter(x_positions[0], y_positions[0], color='green', s=80, marker='o')
    ax_inset.scatter(x_positions[-1], y_positions[-1], color='red', s=80, marker='o')
    
    # Add an arrow to show direction of movement
    arrow_idx = len(distances) // 2
    ax_inset.arrow(x_positions[arrow_idx], y_positions[arrow_idx], 
                  (x_positions[arrow_idx+1] - x_positions[arrow_idx])*3, 
                  (y_positions[arrow_idx+1] - y_positions[arrow_idx])*3,
                  head_width=0.1, head_length=0.15, fc='red', ec='red')
    
    # Draw room boundary if room acoustics is enabled
    if ENABLE_ROOM_ACOUSTICS:
        circle = plt.Circle((0, 0), ROOM_RADIUS, fill=False, linestyle='--', color='gray', alpha=0.5)
        ax_inset.add_artist(circle)
        ax_inset.text(0, ROOM_RADIUS, "Room", ha='center', va='bottom', 
                     color='gray', fontsize=8, alpha=0.7)
    
    # Set equal aspect ratio for inset
    ax_inset.set_aspect('equal')
    ax_inset.set_title("Top-Down View", fontsize=10)
    ax_inset.grid(True, linestyle='--', alpha=0.3)
    ax_inset.set_xlabel("X (meters)")
    ax_inset.set_ylabel("Y (meters)")
    
    # Set axis limits to show the room
    max_range = max(abs(INITIAL_DISTANCE), abs(FINAL_DISTANCE), ROOM_RADIUS) * 1.2
    ax_inset.set_xlim(-max_range, max_range)
    ax_inset.set_ylim(-max_range, max_range)
    
    # Save figure
    plt.savefig(output_path, dpi=300, bbox_inches='tight')
    plt.close()
    
    print(f"Saved trajectory plot to: {output_path}")

def main():
    """Generate approaching sound from different directions using CTF SOFA models"""
    if not SOFA_AVAILABLE:
        print("ERROR: sofar library is required. Please install it with:")
        print("pip install sofar")
        return
    
    # Set up the room model if room acoustics is enabled
    room = None
    if ENABLE_ROOM_ACOUSTICS and PYROOMACOUSTICS_AVAILABLE:
        room = setup_room_model()
    
    # Generate source signal (stereo)
    print("Generating source signal...")
    mono_source = generate_pink_noise(DURATION, SAMPLE_RATE)
    source_signal = np.column_stack((mono_source, mono_source))
    
    # Process each SOFA file
    valid_files = []
    
    for sofa_path in sofa_files:
        if not os.path.exists(sofa_path):
            print(f"File not found: {sofa_path}")
            continue
        
        valid_files.append(sofa_path)
        file_name = os.path.basename(sofa_path).replace('.sofa', '')
        print(f"\nProcessing {file_name}...")
        
        try:
            # Load SOFA file
            hrtf, dimensions = load_sofa_file(sofa_path)
            
            if hrtf is None:
                print(f"Failed to load {file_name}, skipping.")
                continue
            
            # Create approaching movement for each azimuth
            for azimuth in AZIMUTH_ANGLES:
                print(f"\nGenerating approaching sound from azimuth {azimuth}°...")
                
                # Create approaching movement with enhanced effects
                output_signal, distances = create_approaching_movement(hrtf, source_signal, azimuth, room)
                
                # Save audio file
                timestamp = time.strftime("%Y%m%d_%H%M%S")
                
                # Get direction string
                if azimuth == 0:
                    direction = "front"
                elif azimuth == 90:
                    direction = "right"
                elif azimuth == -90:
                    direction = "left"
                else:
                    direction = f"az{azimuth}"
                
                # Create effects string
                effect_indicators = []
                if ENABLE_ROOM_ACOUSTICS: effect_indicators.append("room")
                if ENABLE_DOPPLER: effect_indicators.append("doppler")
                if ENABLE_LOOMING: effect_indicators.append("looming")
                effects_str = "_".join(effect_indicators) if effect_indicators else "basic"
                
                output_filename = f"approaching_{direction}_{file_name}_{effects_str}.wav"
                output_path = os.path.join(output_dir, output_filename)
                sf.write(output_path, output_signal, SAMPLE_RATE)
                print(f"Saved audio to: {output_path}")
                
                # Create trajectory plot for each azimuth
                plot_path = os.path.join(output_dir, 
                                      f"trajectory_{direction}_{file_name}.png")
                plot_distance_trajectory(distances, azimuth, plot_path)
            
        except Exception as e:
            print(f"Error processing {file_name}: {e}")
            import traceback
            traceback.print_exc()
    
    print("\nApproaching movement processing complete!")
    print(f"Total SOFA models processed: {len(valid_files)}")
    print(f"All files saved to: {output_dir}")
    
    # Summarize settings used
    print("\nSettings used:")
    print(f"  Initial distance: {INITIAL_DISTANCE}m")
    print(f"  Final distance: {FINAL_DISTANCE}m")
    print(f"  Motion duration: {DURATION}s")
    print(f"  Azimuths: {AZIMUTH_ANGLES}")
    print(f"  Elevation: {CONSTANT_ELEVATION}°")
    print(f"  Fade-out duration: {FADE_OUT_DURATION}s")
    print(f"  Room acoustics: {'Enabled' if ENABLE_ROOM_ACOUSTICS else 'Disabled'}")
    if ENABLE_ROOM_ACOUSTICS:
        print(f"    Room radius: {ROOM_RADIUS}m")
        print(f"    RT60: {RT60}s")
    print(f"  Doppler effect: {'Enabled' if ENABLE_DOPPLER else 'Disabled'}")
    print(f"  Looming effect: {'Enabled' if ENABLE_LOOMING else 'Disabled'}")

if __name__ == "__main__":
    main()
```

```python
"""
Enhanced SOFA HRTF Distance-Based Movement

This script creates binaural audio files where a sound source approaches the listener
at ear level (0° elevation), maintaining the same angle but decreasing in distance,
for various HRTF models. This version creates variants with different azimuth angles.

The script includes room acoustics, looming effect, and Doppler effect to create
a more realistic approaching sound with smooth transitions.
"""
import numpy as np
import scipy.signal as signal
import soundfile as sf
import matplotlib.pyplot as plt
import os
import time
import datetime

# First, ensure required libraries are installed
try:
    import sofar
    SOFA_AVAILABLE = True
except ImportError:
    print("Warning: sofar library not found. Please install with: pip install sofar")
    SOFA_AVAILABLE = False

try:
    import pyroomacoustics as pra
    PYROOMACOUSTICS_AVAILABLE = True
except ImportError:
    print("Warning: pyroomacoustics library not found. Please install with: pip install pyroomacoustics")
    PYROOMACOUSTICS_AVAILABLE = False

#############################################################
# CONTROL PARAMETERS - ADJUST THESE TO CHANGE THE SIMULATION
#############################################################

# Spatial parameters
INITIAL_DISTANCE = 2.0    # Starting distance in meters
FINAL_DISTANCE = 0.3      # Final distance in meters (closest approach)
AZIMUTH_ANGLES = [90, 0, -90]  # Angles to generate (right, front, left)
CONSTANT_ELEVATION = 0    # Fixed elevation (ear level)

# Time parameters
DURATION = 3.0            # Duration of movement in seconds (longer = slower movement)
SAMPLE_RATE = 48000       # Audio sample rate in Hz
N_SEGMENTS = 120           # Number of segments for processing (higher = smoother movement)
FADE_OUT_DURATION = 0   # Duration of the fade-out at the end (seconds)

# Room acoustics parameters
ROOM_RADIUS = 2.0         # Radius of the spherical room in meters
WALL_ABSORPTION = 0.3     # Wall absorption coefficient (0-1, higher = more absorption)
RT60 = 0.4                # Desired RT60 reverberation time (seconds)

# Effect toggles
ENABLE_ROOM_ACOUSTICS = True   # Whether to simulate room acoustics
ENABLE_DOPPLER = True          # Whether to simulate Doppler effect 
ENABLE_LOOMING = True          # Whether to enhance approaching sound with looming effect

# Use all SOFA files
sofa_files = [
    r"C:\Users\cogpsy-vrlab\Downloads\HRIRs_neutral_head_orientation_v4\SOFA\FABIAN_HRIR_measured_HATO_0.sofa",
    r"C:\Users\cogpsy-vrlab\Downloads\HRIRs_neutral_head_orientation_v4\SOFA\FABIAN_HRIR_modeled_HATO_0.sofa",
    r"C:\Users\cogpsy-vrlab\Downloads\HRIRs_neutral_head_orientation_v4\SOFA\FABIAN_CTF_measured.sofa",
    r"C:\Users\cogpsy-vrlab\Downloads\HRIRs_neutral_head_orientation_v4\SOFA\FABIAN_CTF_measured_smoothed.sofa"
]

# Create output directory with timestamp
timestamp = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
output_dir = os.path.join(r"C:\Users\cogpsy-vrlab\Documents\PPS_module\BreathingPilot\LoomingStimuli", f"looming_stimuli_{timestamp}")
os.makedirs(output_dir, exist_ok=True)

# Calculate speed of sound (m/s)
SPEED_OF_SOUND = 343.0

def generate_pink_noise(duration, fs):
    """Generate pink noise using FFT-based method"""
    n_samples = int(duration * fs)
    
    # Generate white noise
    np.random.seed(42)  # For reproducibility
    white_noise = np.random.randn(n_samples)
    
    # Get the FFT of the white noise
    X = np.fft.rfft(white_noise)
    
    # Create pink filter (1/f)
    n_bins = len(X)
    f = np.linspace(1, n_bins, n_bins)
    f[0] = 1  # Avoid division by zero
    
    # Apply 1/sqrt(f) filter for power spectrum ~ 1/f
    X = X / np.sqrt(f)
    
    # Convert back to time domain
    pink_noise = np.real(np.fft.irfft(X, n_samples))
    
    # Normalize
    pink_noise = pink_noise / np.max(np.abs(pink_noise)) * 0.8
    
    # Apply fade in/out to avoid clicks
    fade_samples = int(0.05 * fs)  # 50ms fade
    fade_in = np.linspace(0, 1, fade_samples)
    fade_out = np.linspace(1, 0, fade_samples)
    
    pink_noise[:fade_samples] *= fade_in
    pink_noise[-fade_samples:] *= fade_out
    
    return pink_noise

def load_sofa_file(sofa_path):
    """Load HRTF data from SOFA file"""
    try:
        print(f"Loading SOFA file: {os.path.basename(sofa_path)}")
        
        # Load the SOFA file
        hrtf = sofar.read_sofa(sofa_path)
        
        # Display basic SOFA info
        print(f"SOFA convention: {hrtf.GLOBAL_SOFAConventions} v{hrtf.GLOBAL_SOFAConventionsVersion}")
        
        # Get dimensions
        dimensions = {}
        for dim_name in ['M', 'R', 'N']:
            try:
                dimensions[dim_name] = hrtf.get_dimension(dim_name)
                print(f"{dim_name} = {dimensions[dim_name]}")
            except:
                dimensions[dim_name] = None
                print(f"{dim_name} = Unknown")
        
        return hrtf, dimensions
        
    except Exception as e:
        print(f"Error loading SOFA file: {e}")
        import traceback
        traceback.print_exc()
        return None, None

def find_nearest_hrtf(hrtf, target_azimuth, target_elevation=0, target_distance=1.0):
    """Find the measurement closest to the target azimuth, elevation, and distance"""
    try:
        source_positions = np.array(hrtf.SourcePosition)
        
        # Handle different position array shapes
        if source_positions.shape[0] == 3 and len(source_positions.shape) == 2:
            source_positions = source_positions.T
        
        # Check if positions are in degrees or radians
        position_type = hrtf.SourcePosition_Type
        max_azimuth = np.max(np.abs(source_positions[:, 0]))
        positions_in_degrees = max_azimuth > 6.28
        
        if positions_in_degrees:
            # Both source positions and target are in degrees
            az_compare = target_azimuth
            el_compare = target_elevation
        else:
            # Source positions in radians, convert target to radians
            az_compare = np.deg2rad(target_azimuth)
            el_compare = np.deg2rad(target_elevation)
        
        # Handle negative azimuths if data is in 0-360 format
        if positions_in_degrees and target_azimuth < 0:
            # Check if we have negative azimuths in the data
            has_negative_az = np.any(source_positions[:, 0] < 0)
            if not has_negative_az:
                # Convert target from -180 to 180 format to 0 to 360 format
                az_compare = (target_azimuth + 360) % 360
        
        # Create distance metric
        # For spherical coordinates: focus mostly on azimuth, then elevation
        # Ignore distance since most HRTF databases only have one distance
        if position_type == 'spherical':
            dists = np.sqrt(
                2 * (source_positions[:, 0] - az_compare)**2 + 
                (source_positions[:, 1] - el_compare)**2
            )
        else:
            # For Cartesian: calculate actual 3D distance
            # Create a target position in Cartesian
            if position_type == 'cartesian':
                # Convert spherical to Cartesian
                target_x = target_distance * np.cos(np.deg2rad(target_elevation)) * np.cos(np.deg2rad(target_azimuth))
                target_y = target_distance * np.cos(np.deg2rad(target_elevation)) * np.sin(np.deg2rad(target_azimuth))
                target_z = target_distance * np.sin(np.deg2rad(target_elevation))
                target_pos = np.array([target_x, target_y, target_z])
                
                # Calculate Euclidean distance to each position
                dists = np.sqrt(np.sum((source_positions - target_pos)**2, axis=1))
            else:
                # Unknown position type, use simple difference
                dists = np.abs(source_positions[:, 0] - az_compare)
                
        # Find nearest position
        nearest_idx = np.argmin(dists)
        nearest_pos = source_positions[nearest_idx]
        
        return nearest_idx, nearest_pos
        
    except Exception as e:
        print(f"Error finding nearest HRTF: {e}")
        import traceback
        traceback.print_exc()
        return 0, None

def extract_hrtf_data(hrtf, idx):
    """Extract HRTF impulse responses for a given index"""
    hrtf_data = hrtf.Data_IR
    
    # Check dimensions and extract
    if hrtf_data.shape[0] == 2 and hrtf_data.shape[1] > 2:
        # Shape is (R, M, N)
        hrtf_left = hrtf_data[0, idx, :]
        hrtf_right = hrtf_data[1, idx, :]
    else:
        # Shape is (M, R, N)
        try:
            hrtf_left = hrtf_data[idx, 0, :]
            hrtf_right = hrtf_data[idx, 1, :]
        except:
            # Try other dimension orderings if the above fails
            try:
                if len(hrtf_data.shape) == 4:  # If 4D array
                    hrtf_left = hrtf_data[idx, 0, 0, :]
                    hrtf_right = hrtf_data[idx, 1, 0, :]
                else:
                    raise ValueError(f"Unexpected HRTF data shape: {hrtf_data.shape}")
            except Exception as e:
                print(f"Error extracting HRTF data: {e}")
                # Return empty IRs as fallback
                dummy_ir = np.zeros(256)
                dummy_ir[0] = 1.0
                return dummy_ir, dummy_ir
    
    return hrtf_left, hrtf_right

def setup_room_model():
    """Create a spherical room model using pyroomacoustics"""
    if not PYROOMACOUSTICS_AVAILABLE or not ENABLE_ROOM_ACOUSTICS:
        print("Room acoustics disabled or pyroomacoustics not available")
        return None
    
    try:
        # Calculate actual RT60 based on room parameters
        # Using Sabine's formula: RT60 = 0.161 * V / (A * S)
        # where V is volume, A is absorption coefficient, S is surface area
        room_volume = (4/3) * np.pi * ROOM_RADIUS**3
        room_surface = 4 * np.pi * ROOM_RADIUS**2
        
        # Adjust absorption to match desired RT60 if possible
        sabine_absorption = 0.161 * room_volume / (RT60 * room_surface)
        
        # Clamp absorption between 0.01 and 0.99
        absorption = min(0.99, max(0.01, sabine_absorption))
        
        print(f"Creating spherical room model with radius {ROOM_RADIUS}m")
        print(f"Calculated absorption coefficient: {absorption:.2f} for RT60 of {RT60}s")
        
        # Create a circular room approximation (2D projection of sphere)
        num_walls = 16
        angles = np.linspace(0, 2*np.pi, num_walls, endpoint=False)
        corners = np.array([[ROOM_RADIUS * np.cos(ang), ROOM_RADIUS * np.sin(ang)] 
                          for ang in angles]).T
        
        # Create room with the specified absorption
        materials = pra.Material(energy_absorption=absorption)
        room = pra.Room.from_corners(corners, fs=SAMPLE_RATE, materials=materials, max_order=3)
        
        # Create stereo microphone array (listener's ears)
        ear_spacing = 0.18  # typical head size
        mic_positions = np.array([
            [-ear_spacing/2, 0],  # Left ear
            [ear_spacing/2, 0]    # Right ear
        ]).T
        
        mic_array = pra.MicrophoneArray(mic_positions, room.fs)
        room.add_microphone_array(mic_array)
        
        return room
    
    except Exception as e:
        print(f"Error setting up room model: {e}")
        import traceback
        traceback.print_exc()
        return None

def apply_room_acoustics(audio, source_position, room):
    """Apply room acoustics to the audio using the pyroomacoustics model"""
    if room is None:
        return audio
    
    try:
        # Reset any existing sources
        room.sources = []
        
        # Add new source with the given position and mono audio
        # For simplicity, use the average of stereo channels
        if audio.ndim == 2:
            mono_signal = np.mean(audio, axis=1)
        else:
            mono_signal = audio
        
        # Add the source to the room
        room.add_source(source_position, signal=mono_signal)
        
        # Compute the room impulse responses
        room.compute_rir()
        
        # Simulate the room acoustics
        room_output = room.simulate()
        
        # Convert to stereo output
        output_signal = np.zeros((len(room_output), 2))
        output_signal[:, 0] = room_output[:, 0]  # Left ear
        output_signal[:, 1] = room_output[:, 1]  # Right ear
        
        # Normalize output
        max_val = np.max(np.abs(output_signal))
        if max_val > 0:
            output_signal = output_signal / max_val * 0.9
            
        return output_signal
        
    except Exception as e:
        print(f"Error applying room acoustics: {e}")
        import traceback
        traceback.print_exc()
        return audio

def apply_basic_distance_model(audio, distance, reference_distance=1.0, final_distance=FINAL_DISTANCE):
    """
    Apply a basic distance model to simulate source distance with improved transition at close distances
    
    This uses a modified inverse square law for amplitude and
    adds a low-pass filter effect for further distances
    """
    # Apply a modified inverse square law for amplitude that limits extreme amplification at close distances
    distance_ratio = max(0.1, min(distance / reference_distance, 10.0))
    
    # Modified amplitude scaling that softens the transition as we get very close
    # Use a nonlinear curve that asymptotically approaches a maximum value
    if distance <= final_distance * 1.5:
        # For distances very close to the final distance, use a gentler curve
        min_ratio = final_distance / reference_distance
        normalized_ratio = (distance_ratio - min_ratio) / (1.0 - min_ratio)
        # Use a sigmoid-like function to smooth the transition
        transition_factor = 0.5 * (1 - np.cos(normalized_ratio * np.pi))
        amplitude_scale = 1.0 / (distance_ratio ** (1.5 * transition_factor))
    else:
        # Standard inverse square law for further distances
        amplitude_scale = 1.0 / (distance_ratio ** 2)
    
    # Normalize scale to avoid extreme amplification
    amplitude_scale = min(amplitude_scale, 2.5)
    distance_audio = audio * amplitude_scale
    
    # Apply distance-based filtering (closer = brighter, further = more muffled)
    if distance > reference_distance:
        # Create a lowpass filter for distant sounds
        cutoff_freq = max(1000, min(20000, 20000 / distance_ratio))
        cutoff_norm = cutoff_freq / (SAMPLE_RATE/2)  # Normalize to Nyquist
        b, a = signal.butter(2, cutoff_norm, btype='low')
        
        # Apply filter to both channels
        distance_audio[:, 0] = signal.filtfilt(b, a, distance_audio[:, 0])
        distance_audio[:, 1] = signal.filtfilt(b, a, distance_audio[:, 1])
    
    return distance_audio

def apply_looming_effect(audio, distance, max_distance):
    """
    Apply looming effect to enhance the perception of an approaching sound
    
    Looming involves spectral and amplitude changes that create a sense
    of urgency as objects approach
    """
    if not ENABLE_LOOMING:
        return audio
    
    # Normalize distance to 0-1 range (1 = furthest, 0 = closest)
    norm_distance = distance / max_distance
    
    # Create enhanced audio
    looming_audio = audio.copy()
    
    # 1. Enhanced amplitude changes (nonlinear)
    # Approach sounds are perceived as louder than expected from inverse square law
    # Using a gentler curve for close distances to prevent the jump
    if norm_distance < 0.2:
        # Softer enhancement for very close distances
        looming_factor = 1.0 + (1.0 - norm_distance) * 0.5
    else:
        looming_factor = 1.0 + (1.0 - norm_distance) ** 1.5
    
    looming_audio *= looming_factor
    
    # 2. Spectral emphasis changes with distance
    # As objects approach, mid-high frequencies become more prominent
    if norm_distance < 0.8:  # Only apply for closer sounds
        # Boost mid and high frequencies as object gets closer
        # Use a custom method instead of highshelf filter
        boost_amount = (1.0 - norm_distance) * 5.0  # dB boost (max 5dB, reduced from 6dB)
        shelf_freq = 1000  # Hz
        
        # Simple frequency boost using FFT
        for ch in range(looming_audio.shape[1]):
            # Get channel data
            channel_data = looming_audio[:, ch]
            
            # Apply FFT
            fft_data = np.fft.rfft(channel_data)
            freqs = np.fft.rfftfreq(len(channel_data), 1.0/SAMPLE_RATE)
            
            # Create frequency-dependent gain
            gain_factor = np.ones(len(freqs))
            
            # Apply boost to frequencies above shelf_freq
            above_shelf = freqs >= shelf_freq
            # Convert dB to amplitude ratio (dB = 20 * log10(ratio))
            boost_ratio = 10 ** (boost_amount / 20)
            
            # Create a gradual transition
            transition_width = 200  # Hz
            for i, freq in enumerate(freqs):
                if freq >= shelf_freq:
                    # Linear transition from 1.0 to boost_ratio over transition_width
                    if freq < shelf_freq + transition_width:
                        transition_factor = (freq - shelf_freq) / transition_width
                        gain_factor[i] = 1.0 + (boost_ratio - 1.0) * transition_factor
                    else:
                        gain_factor[i] = boost_ratio
            
            # Apply gain
            fft_data *= gain_factor
            
            # Apply inverse FFT
            looming_audio[:, ch] = np.real(np.fft.irfft(fft_data, len(channel_data)))
    
    # 3. Add slight harmonic distortion for very close sounds (creates urgency)
    if norm_distance < 0.3:
        # Reduce distortion for very close distances to avoid the jump at the end
        if norm_distance < 0.1:
            # Gradually reduce distortion as we get extremely close
            distortion_amount = (0.1 - norm_distance) * 0.03  # Much gentler distortion
        else:
            distortion_amount = (0.3 - norm_distance) * 0.07  # Reduced from 0.1
        
        # Simple waveshaping distortion
        looming_audio = np.tanh(looming_audio * (1.0 + distortion_amount)) / (1.0 + distortion_amount)
    
    # Normalize to avoid clipping
    max_val = np.max(np.abs(looming_audio))
    if max_val > 1.0:
        looming_audio /= max_val
    
    return looming_audio

def apply_doppler_effect(source_signal, distances, fs=SAMPLE_RATE):
    """
    Apply Doppler effect to the audio based on the source's velocity
    
    The Doppler effect causes frequency shifts based on relative velocity:
    - Approaching sound sources have higher pitch
    - Receding sound sources have lower pitch
    """
    if not ENABLE_DOPPLER or len(distances) < 2:
        return source_signal
    
    try:
        # Calculate time step
        time_step = DURATION / len(distances)
        
        # Calculate velocities (m/s) - negative velocity means approaching
        velocities = np.zeros(len(distances) - 1)
        for i in range(len(distances) - 1):
            velocities[i] = (distances[i+1] - distances[i]) / time_step
        
        # Add a final velocity (repeat the last one)
        velocities = np.append(velocities, velocities[-1])
        
        # Smooth velocities to avoid abrupt changes, especially at the end
        window_size = 5
        padded_velocities = np.pad(velocities, (window_size//2, window_size//2), mode='edge')
        smoothed_velocities = np.zeros_like(velocities)
        
        for i in range(len(velocities)):
            smoothed_velocities[i] = np.mean(padded_velocities[i:i+window_size])
        
        velocities = smoothed_velocities
        
        # Calculate the Doppler shift factor: f' = f * (v_sound / (v_sound + v_source))
        # Note: approaching source has negative velocity, so denominator decreases
        doppler_factors = SPEED_OF_SOUND / (SPEED_OF_SOUND + velocities)
        
        # Limit the maximum frequency shift to avoid extreme effects
        doppler_factors = np.clip(doppler_factors, 0.5, 2.0)
        
        # Get length of source signal
        n_samples = len(source_signal)
        
        # Initialize output signal
        output_signal = np.zeros_like(source_signal)
        
        # Calculate resampling indices for each sample
        # This maps where each output sample should come from in the input
        resample_indices = np.zeros(n_samples)
        current_index = 0.0
        
        segment_length = n_samples // len(doppler_factors)
        
        for i in range(n_samples):
            # Determine which segment we're in
            segment_idx = min(i // segment_length, len(doppler_factors) - 1)
            
            # Get the doppler factor for this segment
            doppler_factor = doppler_factors[segment_idx]
            
            # Update the current index
            resample_indices[i] = current_index
            current_index += doppler_factor
        
        # Clip indices to valid range
        max_idx = n_samples - 1
        resample_indices = np.clip(resample_indices, 0, max_idx)
        
        # Create interpolated output using linear interpolation
        for ch in range(source_signal.shape[1]):
            output_signal[:, ch] = np.interp(
                resample_indices, 
                np.arange(n_samples), 
                source_signal[:, ch]
            )
        
        return output_signal
        
    except Exception as e:
        print(f"Error applying Doppler effect: {e}")
        import traceback
        traceback.print_exc()
        return source_signal

def create_linear_distance_trajectory():
    """
    Create a linear distance trajectory with constant approach rate
    """
    # Create a basic linear trajectory from initial to final distance
    distances = np.linspace(INITIAL_DISTANCE, FINAL_DISTANCE, N_SEGMENTS)
    
    return distances

def create_approaching_movement(hrtf, source_signal, azimuth, room=None, verbose=True):
    """
    Create audio where the sound stays at constant azimuth
    but approaches the listener from far to near
    """
    # Calculate total samples
    total_samples = len(source_signal)
    segment_samples = total_samples // N_SEGMENTS
    
    # Create linear distance trajectory (constant approach rate)
    distances = create_linear_distance_trajectory()
    
    # Create output buffer
    output_signal = np.zeros((total_samples, 2))
    
    # We'll use the same HRTF for all segments (constant angle)
    # but will apply distance-based effects to simulate the approach
    
    # Find the HRTF for our target angle
    nearest_idx, nearest_pos = find_nearest_hrtf(hrtf, azimuth, CONSTANT_ELEVATION)
    hrtf_left, hrtf_right = extract_hrtf_data(hrtf, nearest_idx)
    
    if verbose:
        if nearest_pos is not None:
            if hrtf.SourcePosition_Type == 'spherical' and np.max(np.abs(nearest_pos[0])) > 6.28:
                pos_str = f"({nearest_pos[0]:.1f}°, {nearest_pos[1]:.1f}°)"
            elif hrtf.SourcePosition_Type == 'spherical':
                pos_str = f"({np.rad2deg(nearest_pos[0]):.1f}°, {np.rad2deg(nearest_pos[1]):.1f}°)"
            else:
                pos_str = f"[{nearest_pos[0]:.2f}, {nearest_pos[1]:.2f}, {nearest_pos[2]:.2f}]"
        else:
            pos_str = "unknown"
            
        print(f"Using HRTF measurement at position {pos_str} for all segments")
    
    # Apply the HRTF to the entire source signal
    print("Applying HRTF filter...")
    full_left = signal.fftconvolve(source_signal[:, 0], hrtf_left, mode='same')
    full_right = signal.fftconvolve(source_signal[:, 0], hrtf_right, mode='same')
    binaural_base = np.column_stack((full_left, full_right))
    
    # Apply Doppler effect to the entire signal before segmenting
    if ENABLE_DOPPLER:
        print("Applying Doppler effect...")
        binaural_base = apply_doppler_effect(binaural_base, distances)
    
    # Now process each segment with the appropriate distance model
    if verbose:
        print(f"Processing {N_SEGMENTS} segments with distance effects...")
    
    for i in range(N_SEGMENTS):
        # Calculate segment boundaries
        start_idx = i * segment_samples
        end_idx = min((i + 1) * segment_samples, total_samples)
        
        # Get the binaural segment
        segment = binaural_base[start_idx:end_idx]
        
        # Get distance for this segment
        distance = distances[i]
        
        # 1. Apply basic distance model (amplitude and filtering)
        processed_segment = apply_basic_distance_model(segment, distance, final_distance=FINAL_DISTANCE)
        
        # 2. Apply looming effect if enabled
        if ENABLE_LOOMING:
            processed_segment = apply_looming_effect(processed_segment, distance, INITIAL_DISTANCE)
        
        # 3. Apply room acoustics if enabled
        if ENABLE_ROOM_ACOUSTICS and room is not None:
            # Convert spherical coordinates to Cartesian for room model
            source_position = [distance * np.cos(np.deg2rad(azimuth + 90)), 
                             distance * np.sin(np.deg2rad(azimuth + 90))]
            
            # Process with room acoustics
            # Note: We'll skip this for now and apply room acoustics only to final output
            # This is because applying room acoustics to each segment is computationally intensive
            pass
        
        # Add to output buffer
        output_signal[start_idx:end_idx] = processed_segment
        
        # Print progress
        if verbose and (i % 10 == 0 or i == N_SEGMENTS - 1):
            print(f"Processed segment {i+1}/{N_SEGMENTS} at distance {distance:.2f}m")
    
    # Apply a gentle smoothing between segments to avoid clicks
    if N_SEGMENTS > 1:
        print("Applying smoothing between segments...")
        
        # Use a short crossfade at segment boundaries
        crossfade_len = min(segment_samples // 10, 200)
        
        for i in range(1, N_SEGMENTS):
            boundary_idx = i * segment_samples
            
            # Make sure we have enough samples for crossfade
            if boundary_idx - crossfade_len // 2 >= 0 and boundary_idx + crossfade_len // 2 < total_samples:
                # Get the audio around the boundary
                start_idx = boundary_idx - crossfade_len // 2
                end_idx = boundary_idx + crossfade_len // 2
                
                # Ensure we have an even length for the window
                window_len = end_idx - start_idx
                
                # Get the boundary audio segment
                boundary_audio = output_signal[start_idx:end_idx]
                
                # Create a window of exactly the right length
                window = np.hanning(window_len)
                
                # Apply window to both channels
                boundary_audio *= window[:, np.newaxis]
                
                # Put back into the output
                output_signal[start_idx:end_idx] = boundary_audio
    
    # Apply room acoustics to the final output if enabled
    if ENABLE_ROOM_ACOUSTICS and room is not None and PYROOMACOUSTICS_AVAILABLE:
        print("Applying room acoustics to final output...")
        # For demo purposes, use the middle distance position
        mid_distance = (INITIAL_DISTANCE + FINAL_DISTANCE) / 2
        source_position = [mid_distance * np.cos(np.deg2rad(azimuth + 90)), 
                         mid_distance * np.sin(np.deg2rad(azimuth + 90))]
        
        # Apply room acoustics
        try:
            output_signal = apply_room_acoustics(output_signal, source_position, room)
            print("Room acoustics applied successfully")
        except Exception as e:
            print(f"Error applying room acoustics: {e}")
    
    # Apply a final fade-out at the end to ensure smooth termination
    fade_samples = int(FADE_OUT_DURATION * SAMPLE_RATE)
    fade_start = total_samples - fade_samples
    
    # Create fade-out window
    fade_window = np.ones(total_samples)
    fade_window[fade_start:] = np.linspace(1.0, 0.0, fade_samples)
    
    # Apply fade-out to both channels
    output_signal[:, 0] *= fade_window
    output_signal[:, 1] *= fade_window
    
    # Normalize output
    max_val = np.max(np.abs(output_signal))
    if max_val > 0:
        output_signal = output_signal / max_val * 0.9
    
    return output_signal, distances

def plot_distance_trajectory(distances, azimuth, output_path):
    """Plot the trajectory of the approaching sound (distance over time)"""
    # Create figure
    plt.figure(figsize=(10, 6))
    
    # Define time axis (in seconds)
    time_axis = np.linspace(0, DURATION, len(distances))
    
    # Plot distance vs time
    plt.plot(time_axis, distances, 'b-', linewidth=3)
    plt.scatter(time_axis, distances, c=distances, cmap='viridis', s=80, alpha=0.7)
    
    # Mark start and end points
    plt.scatter(time_axis[0], distances[0], color='green', s=120, marker='o', label="Start (Far)")
    plt.scatter(time_axis[-1], distances[-1], color='red', s=120, marker='o', label="End (Near)")
    
    # Add grid and labels
    plt.grid(True, linestyle='--', alpha=0.5)
    
    # Create azimuth direction string
    if azimuth == 0:
        direction = "Front"
    elif azimuth == 90:
        direction = "Right"
    elif azimuth == -90:
        direction = "Left"
    else:
        direction = f"{azimuth}°"
    
    plt.title(f"Sound Source Approaching from the {direction} ({azimuth}° azimuth)", fontsize=14)
    plt.xlabel("Time (seconds)", fontsize=12)
    plt.ylabel("Distance from Listener (meters)", fontsize=12)
    plt.legend(fontsize=10)
    
    # Add annotations
    plt.annotate("Far", (time_axis[0], distances[0]), 
                xytext=(time_axis[0]-0.2, distances[0]+0.2),
                fontsize=12, fontweight='bold')
    plt.annotate("Near", (time_axis[-1], distances[-1]), 
                xytext=(time_axis[-1]-0.2, distances[-1]+0.2),
                fontsize=12, fontweight='bold')
    
    # Add top-down view of the scene
    ax_inset = plt.axes([0.65, 0.15, 0.3, 0.3])  # [left, bottom, width, height]
    
    # Plot listener (center) and sound positions
    ax_inset.scatter([0], [0], color='black', s=200, marker='o', label="Listener")
    
    # Calculate positions along azimuth at different distances
    # We rotate by 90° to convert from azimuth to Cartesian
    x_positions = distances * np.cos(np.deg2rad(azimuth + 90))
    y_positions = distances * np.sin(np.deg2rad(azimuth + 90))
    
    # Plot trajectory points
    ax_inset.scatter(x_positions, y_positions, c=np.arange(len(distances)), cmap='viridis', s=50, alpha=0.7)
    ax_inset.plot(x_positions, y_positions, 'k-', alpha=0.5)
    
    # Mark start and end
    ax_inset.scatter(x_positions[0], y_positions[0], color='green', s=80, marker='o')
    ax_inset.scatter(x_positions[-1], y_positions[-1], color='red', s=80, marker='o')
    
    # Add an arrow to show direction of movement
    arrow_idx = len(distances) // 2
    ax_inset.arrow(x_positions[arrow_idx], y_positions[arrow_idx], 
                  (x_positions[arrow_idx+1] - x_positions[arrow_idx])*3, 
                  (y_positions[arrow_idx+1] - y_positions[arrow_idx])*3,
                  head_width=0.1, head_length=0.15, fc='red', ec='red')
    
    # Draw room boundary if room acoustics is enabled
    if ENABLE_ROOM_ACOUSTICS:
        circle = plt.Circle((0, 0), ROOM_RADIUS, fill=False, linestyle='--', color='gray', alpha=0.5)
        ax_inset.add_artist(circle)
        ax_inset.text(0, ROOM_RADIUS, "Room", ha='center', va='bottom', 
                     color='gray', fontsize=8, alpha=0.7)
    
    # Set equal aspect ratio for inset
    ax_inset.set_aspect('equal')
    ax_inset.set_title("Top-Down View", fontsize=10)
    ax_inset.grid(True, linestyle='--', alpha=0.3)
    ax_inset.set_xlabel("X (meters)")
    ax_inset.set_ylabel("Y (meters)")
    
    # Set axis limits to show the room
    max_range = max(abs(INITIAL_DISTANCE), abs(FINAL_DISTANCE), ROOM_RADIUS) * 1.2
    ax_inset.set_xlim(-max_range, max_range)
    ax_inset.set_ylim(-max_range, max_range)
    
    # Save figure
    plt.savefig(output_path, dpi=300, bbox_inches='tight')
    plt.close()
    
    print(f"Saved trajectory plot to: {output_path}")

def main():
    """Generate approaching sound from different directions using CTF SOFA models"""
    if not SOFA_AVAILABLE:
        print("ERROR: sofar library is required. Please install it with:")
        print("pip install sofar")
        return
    
    # Set up the room model if room acoustics is enabled
    room = None
    if ENABLE_ROOM_ACOUSTICS and PYROOMACOUSTICS_AVAILABLE:
        room = setup_room_model()
    
    # Generate source signal (stereo)
    print("Generating source signal...")
    mono_source = generate_pink_noise(DURATION, SAMPLE_RATE)
    source_signal = np.column_stack((mono_source, mono_source))
    
    # Process each SOFA file
    valid_files = []
    
    for sofa_path in sofa_files:
        if not os.path.exists(sofa_path):
            print(f"File not found: {sofa_path}")
            continue
        
        valid_files.append(sofa_path)
        file_name = os.path.basename(sofa_path).replace('.sofa', '')
        print(f"\nProcessing {file_name}...")
        
        try:
            # Load SOFA file
            hrtf, dimensions = load_sofa_file(sofa_path)
            
            if hrtf is None:
                print(f"Failed to load {file_name}, skipping.")
                continue
            
            # Create approaching movement for each azimuth
            for azimuth in AZIMUTH_ANGLES:
                print(f"\nGenerating approaching sound from azimuth {azimuth}°...")
                
                # Create approaching movement with enhanced effects
                output_signal, distances = create_approaching_movement(hrtf, source_signal, azimuth, room)
                
                # Save audio file
                timestamp = time.strftime("%Y%m%d_%H%M%S")
                
                # Get direction string
                if azimuth == 0:
                    direction = "front"
                elif azimuth == 90:
                    direction = "right"
                elif azimuth == -90:
                    direction = "left"
                else:
                    direction = f"az{azimuth}"
                
                # Create effects string
                effect_indicators = []
                if ENABLE_ROOM_ACOUSTICS: effect_indicators.append("room")
                if ENABLE_DOPPLER: effect_indicators.append("doppler")
                if ENABLE_LOOMING: effect_indicators.append("looming")
                effects_str = "_".join(effect_indicators) if effect_indicators else "basic"
                
                output_filename = f"approaching_{direction}_az{azimuth}_{file_name}_{effects_str}.wav"
                output_path = os.path.join(output_dir, output_filename)
                sf.write(output_path, output_signal, SAMPLE_RATE)
                print(f"Saved audio to: {output_path}")
                
                # Create trajectory plot for each azimuth
                plot_path = os.path.join(output_dir, 
                                      f"trajectory_{direction}_az{azimuth}_{file_name}.png")
                plot_distance_trajectory(distances, azimuth, plot_path)
            
        except Exception as e:
            print(f"Error processing {file_name}: {e}")
            import traceback
            traceback.print_exc()
    
    print("\nApproaching movement processing complete!")
    print(f"Total SOFA models processed: {len(valid_files)}")
    print(f"All files saved to: {output_dir}")
    
    # Summarize settings used
    print("\nSettings used:")
    print(f"  Initial distance: {INITIAL_DISTANCE}m")
    print(f"  Final distance: {FINAL_DISTANCE}m")
    print(f"  Motion duration: {DURATION}s")
    print(f"  Azimuths: {AZIMUTH_ANGLES}")
    print(f"  Elevation: {CONSTANT_ELEVATION}°")
    print(f"  Fade-out duration: {FADE_OUT_DURATION}s")
    print(f"  Room acoustics: {'Enabled' if ENABLE_ROOM_ACOUSTICS else 'Disabled'}")
    if ENABLE_ROOM_ACOUSTICS:
        print(f"    Room radius: {ROOM_RADIUS}m")
        print(f"    RT60: {RT60}s")
    print(f"  Doppler effect: {'Enabled' if ENABLE_DOPPLER else 'Disabled'}")
    print(f"  Looming effect: {'Enabled' if ENABLE_LOOMING else 'Disabled'}")

if __name__ == "__main__":
    main()
```

```python
"""
SOFA HRTF Moving Sound Source Simulator

This script creates binaural audio files simulating a sound source approaching the listener,
using the Fabian HRTF models for directional cues and a physically-based distance model
inspired by the EPFL LEMA approach.
"""
import numpy as np
import scipy.signal as signal
import soundfile as sf
import matplotlib.pyplot as plt
import os
import time
import datetime

# First, ensure required libraries are installed
try:
    import sofar
    SOFA_AVAILABLE = True
except ImportError:
    print("Warning: sofar library not found. Please install with: pip install sofar")
    SOFA_AVAILABLE = False

#############################################################
# CONTROL PARAMETERS - ADJUST THESE TO CHANGE THE SIMULATION
#############################################################

# Spatial parameters
INITIAL_DISTANCE = 2.0    # Starting distance in meters
FINAL_DISTANCE = 0.5      # Final distance in meters (increased to avoid extreme effects)
AZIMUTH_ANGLES = [90, 0, -90]  # Angles to generate (right, front, left)
CONSTANT_ELEVATION = 0    # Fixed elevation (ear level)
DISTANCE_FACTOR = 40      # Attenuation factor (higher = stronger distance effect)
HEAD_DIAMETER = 0.18      # Diameter of the head in meters

# Time parameters
DURATION = 3.0            # Duration of movement in seconds
SAMPLE_RATE = 44100       # Audio sample rate in Hz
SPEED_OF_SOUND = 343.0    # Speed of sound in m/s

# Use Fabian SOFA files
sofa_files = [
    r"C:\Users\cogpsy-vrlab\Downloads\HRIRs_neutral_head_orientation_v4\SOFA\FABIAN_HRIR_measured_HATO_0.sofa",
    r"C:\Users\cogpsy-vrlab\Downloads\HRIRs_neutral_head_orientation_v4\SOFA\FABIAN_CTF_measured.sofa",
    r"C:\Users\cogpsy-vrlab\Downloads\HRIRs_neutral_head_orientation_v4\SOFA\FABIAN_CTF_measured_smoothed.sofa"
]

# Create output directory with timestamp
timestamp = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
output_dir = os.path.join(r"C:\Users\cogpsy-vrlab\Documents\PPS_module\BreathingPilot\LoomingStimuli", f"looming_stimuli_{timestamp}")
os.makedirs(output_dir, exist_ok=True)

def generate_pink_noise(duration, fs):
    """Generate pink noise using filter-based method (similar to MATLAB code)"""
    # Create filter coefficients for pink noise
    # These are similar to the coefficients used in the MATLAB code
    b = np.array([0.049922035, -0.095993537, 0.050612699, -0.004408786])
    a = np.array([1.0, -2.494956002, 2.017265875, -0.522189400])
    
    # Generate white noise
    np.random.seed(42)  # For reproducibility
    white_noise = np.random.randn(int(duration * fs))
    
    # Apply filter to get pink noise
    pink_noise = signal.lfilter(b, a, white_noise)
    
    # Normalize
    pink_noise = pink_noise / np.max(np.abs(pink_noise)) * 0.8
    
    return pink_noise

def load_sofa_file(sofa_path):
    """Load HRTF data from SOFA file"""
    try:
        print(f"Loading SOFA file: {os.path.basename(sofa_path)}")
        
        # Load the SOFA file
        hrtf = sofar.read_sofa(sofa_path)
        
        # Display basic SOFA info
        print(f"SOFA convention: {hrtf.GLOBAL_SOFAConventions} v{hrtf.GLOBAL_SOFAConventionsVersion}")
        
        # Get dimensions
        dimensions = {}
        for dim_name in ['M', 'R', 'N']:
            try:
                dimensions[dim_name] = hrtf.get_dimension(dim_name)
                print(f"{dim_name} = {dimensions[dim_name]}")
            except:
                dimensions[dim_name] = None
                print(f"{dim_name} = Unknown")
        
        return hrtf, dimensions
        
    except Exception as e:
        print(f"Error loading SOFA file: {e}")
        import traceback
        traceback.print_exc()
        return None, None

def find_nearest_hrtf(hrtf, target_azimuth, target_elevation=0, target_distance=1.0):
    """Find the measurement closest to the target azimuth, elevation, and distance"""
    try:
        source_positions = np.array(hrtf.SourcePosition)
        
        # Handle different position array shapes
        if source_positions.shape[0] == 3 and len(source_positions.shape) == 2:
            source_positions = source_positions.T
        
        # Check if positions are in degrees or radians
        position_type = hrtf.SourcePosition_Type
        max_azimuth = np.max(np.abs(source_positions[:, 0]))
        positions_in_degrees = max_azimuth > 6.28
        
        if positions_in_degrees:
            # Both source positions and target are in degrees
            az_compare = target_azimuth
            el_compare = target_elevation
        else:
            # Source positions in radians, convert target to radians
            az_compare = np.deg2rad(target_azimuth)
            el_compare = np.deg2rad(target_elevation)
        
        # Handle negative azimuths if data is in 0-360 format
        if positions_in_degrees and target_azimuth < 0:
            # Check if we have negative azimuths in the data
            has_negative_az = np.any(source_positions[:, 0] < 0)
            if not has_negative_az:
                # Convert target from -180 to 180 format to 0 to 360 format
                az_compare = (target_azimuth + 360) % 360
        
        # Create distance metric focusing on azimuth and elevation
        if position_type == 'spherical':
            dists = np.sqrt(
                2 * (source_positions[:, 0] - az_compare)**2 + 
                (source_positions[:, 1] - el_compare)**2
            )
        else:
            # For Cartesian or unknown position type
            dists = np.abs(source_positions[:, 0] - az_compare)
                
        # Find nearest position
        nearest_idx = np.argmin(dists)
        nearest_pos = source_positions[nearest_idx]
        
        return nearest_idx, nearest_pos
        
    except Exception as e:
        print(f"Error finding nearest HRTF: {e}")
        import traceback
        traceback.print_exc()
        return 0, None

def extract_hrtf_data(hrtf, idx):
    """Extract HRTF impulse responses for a given index"""
    hrtf_data = hrtf.Data_IR
    
    # Check dimensions and extract
    if hrtf_data.shape[0] == 2 and hrtf_data.shape[1] > 2:
        # Shape is (R, M, N)
        hrtf_left = hrtf_data[0, idx, :]
        hrtf_right = hrtf_data[1, idx, :]
    else:
        # Shape is (M, R, N)
        try:
            hrtf_left = hrtf_data[idx, 0, :]
            hrtf_right = hrtf_data[idx, 1, :]
        except:
            # Try other dimension orderings if the above fails
            try:
                if len(hrtf_data.shape) == 4:  # If 4D array
                    hrtf_left = hrtf_data[idx, 0, 0, :]
                    hrtf_right = hrtf_data[idx, 1, 0, :]
                else:
                    raise ValueError(f"Unexpected HRTF data shape: {hrtf_data.shape}")
            except Exception as e:
                print(f"Error extracting HRTF data: {e}")
                # Return empty IRs as fallback
                dummy_ir = np.zeros(256)
                dummy_ir[0] = 1.0
                return dummy_ir, dummy_ir
    
    return hrtf_left, hrtf_right

def create_approaching_movement_physics(hrtf, source_signal, azimuth, verbose=True):
    """
    Create audio where the sound approaches from a fixed azimuth using the EPFL LEMA
    physically-based approach for distance effects
    """
    # Calculate total samples
    total_samples = int(DURATION * SAMPLE_RATE)
    
    # Use HRTF for directional cues - find closest matching HRTF for our azimuth
    nearest_idx, nearest_pos = find_nearest_hrtf(hrtf, azimuth, CONSTANT_ELEVATION)
    hrtf_left, hrtf_right = extract_hrtf_data(hrtf, nearest_idx)
    
    if verbose:
        if nearest_pos is not None:
            if hrtf.SourcePosition_Type == 'spherical' and np.max(np.abs(nearest_pos[0])) > 6.28:
                pos_str = f"({nearest_pos[0]:.1f}°, {nearest_pos[1]:.1f}°)"
            elif hrtf.SourcePosition_Type == 'spherical':
                pos_str = f"({np.rad2deg(nearest_pos[0]):.1f}°, {np.rad2deg(nearest_pos[1]):.1f}°)"
            else:
                pos_str = f"[{nearest_pos[0]:.2f}, {nearest_pos[1]:.2f}, {nearest_pos[2]:.2f}]"
        else:
            pos_str = "unknown"
            
        print(f"Using HRTF measurement at position {pos_str}")
    
    # Apply the HRTF to the entire source signal for directional cues
    print("Applying HRTF filter for directional cues...")
    dir_left = signal.fftconvolve(source_signal, hrtf_left, mode='same')
    dir_right = signal.fftconvolve(source_signal, hrtf_right, mode='same')
    
    # Now apply the physics-based approach from the MATLAB code for distance effects
    
    # 1. Create source coordinates as it approaches, similar to the MATLAB code
    # Set up the geometry (x, y) coordinates
    # Coordinates of the ears (left and right)
    coo_ear = np.array([[-HEAD_DIAMETER/2, 0], [HEAD_DIAMETER/2, 0]])
    
    # Calculate the y-coordinate based on azimuth
    # For straight approaches, y is constant, x changes
    if azimuth == 0:  # Coming from front
        x_start = -INITIAL_DISTANCE
        x_stop = -FINAL_DISTANCE
        y_value = 0
    elif azimuth == 90:  # Coming from right
        x_start = 0
        x_stop = 0
        y_value = INITIAL_DISTANCE
        y_stop = FINAL_DISTANCE
    elif azimuth == -90:  # Coming from left
        x_start = 0 
        x_stop = 0
        y_value = -INITIAL_DISTANCE
        y_stop = -FINAL_DISTANCE
    else:
        # For other angles, calculate the x and y components
        angle_rad = np.deg2rad(azimuth + 90)  # Convert to radians with 0 = front
        x_start = INITIAL_DISTANCE * np.cos(angle_rad)
        y_start = INITIAL_DISTANCE * np.sin(angle_rad)
        x_stop = FINAL_DISTANCE * np.cos(angle_rad)
        y_stop = FINAL_DISTANCE * np.sin(angle_rad)
    
    # Create source coordinates for each sample
    time_points = np.linspace(0, DURATION, total_samples)
    if azimuth == 90 or azimuth == -90:
        # For right or left approaches, x is constant, y changes
        x_positions = np.full(total_samples, x_start)
        if azimuth == 90:  # Right side
            y_positions = np.linspace(y_value, y_stop, total_samples)
        else:  # Left side
            y_positions = np.linspace(y_value, y_stop, total_samples)
    else:
        # For other angles
        x_positions = np.linspace(x_start, x_stop, total_samples)
        if azimuth == 0:  # Front
            y_positions = np.full(total_samples, y_value)
        else:
            y_positions = np.linspace(y_start, y_stop, total_samples)
    
    # Combine into source coordinates
    source_coordinates = np.column_stack((x_positions, y_positions))
    
    # 2. Calculate distances from the source to each ear at each time point
    # Distance from source to head center
    distance_to_head_center = np.sqrt(source_coordinates[:, 0]**2 + source_coordinates[:, 1]**2)
    
    # Calculate the direction of arrival angle for interaural time difference
    doa = np.zeros(total_samples)
    for i in range(total_samples):
        x_source = source_coordinates[i, 0]
        y_source = source_coordinates[i, 1]
        source_distance = np.sqrt(x_source**2 + y_source**2)
        
        if source_distance > 0:
            # Calculate angle in degrees
            angle_rad = np.arctan2(y_source, x_source)
            doa[i] = np.rad2deg(angle_rad)
        else:
            doa[i] = 0
    
    # Calculate distances from source to each ear
    # Using the spherical head model from the MATLAB code
    # Left ear
    dist_to_left_ear = np.zeros(total_samples)
    # Right ear
    dist_to_right_ear = np.zeros(total_samples)
    
    for i in range(total_samples):
        # Direct Euclidean distance from source to each ear
        x_source = source_coordinates[i, 0]
        y_source = source_coordinates[i, 1]
        
        # Distance to left ear
        dist_to_left_ear[i] = np.sqrt((x_source - coo_ear[0, 0])**2 + (y_source - coo_ear[0, 1])**2)
        
        # Distance to right ear
        dist_to_right_ear[i] = np.sqrt((x_source - coo_ear[1, 0])**2 + (y_source - coo_ear[1, 1])**2)
    
    # 3. Apply amplitude and delay based on distances
    # Calculate amplitude modulation using inverse square law with attenuation factor
    # This is adapted from the MATLAB code's approach
    ref_level = 100  # Reference sound level
    min_distance = 0.001  # To avoid division by zero
    
    # Use the reference amplitude and attenuation factor
    ref_amplitude = 2e-5 * 10**((ref_level-3)/20)
    
    # Calculate amplitude scaling based on distance
    left_amp = np.zeros(total_samples)
    right_amp = np.zeros(total_samples)
    
    for i in range(total_samples):
        # Amplitude based on distance, with directional bias
        d_left = max(dist_to_left_ear[i], min_distance)
        d_right = max(dist_to_right_ear[i], min_distance)
        d_center = max(distance_to_head_center[i], min_distance)
        
        # Calculate amplitude attenuation
        # Directional component based on angle (adapted from MATLAB)
        a_L = 5/180
        b_L = -5
        loss_dB_L = a_L * doa[i] + b_L
        
        a_R = -5/180
        b_R = 0
        loss_dB_R = a_R * doa[i] + b_R
        
        # Combine direction and distance effects
        left_amp[i] = 2e-5 * 10**((loss_dB_L + DISTANCE_FACTOR * np.log10(1/d_center))/20)
        right_amp[i] = 2e-5 * 10**((loss_dB_R + DISTANCE_FACTOR * np.log10(1/d_center))/20)
    
    # Normalize and scale
    left_amp = left_amp / np.max(left_amp) * ref_amplitude
    right_amp = right_amp / np.max(right_amp) * ref_amplitude
    
    # 4. Calculate time delays based on distance differences
    # Convert distance difference to time delay in samples
    left_delay_samples = np.round(dist_to_left_ear / SPEED_OF_SOUND * SAMPLE_RATE).astype(int)
    right_delay_samples = np.round(dist_to_right_ear / SPEED_OF_SOUND * SAMPLE_RATE).astype(int)
    
    # Find maximum delay to ensure we have enough buffer
    max_delay = max(np.max(left_delay_samples), np.max(right_delay_samples))
    
    # 5. Create output buffer with extra space for delays
    output_signal = np.zeros((total_samples + max_delay, 2))
    
    # 6. Apply the amplitude and delay modulation to the directional signal
    for i in range(total_samples):
        # Apply amplitude scaling and add to output with delay
        output_signal[left_delay_samples[i] + i, 0] += dir_left[i] * left_amp[i]
        output_signal[right_delay_samples[i] + i, 1] += dir_right[i] * right_amp[i]
    
    # Ensure we have enough samples and trim if needed
    if len(output_signal) > total_samples:
        output_signal = output_signal[:total_samples]
    
    # 7. Apply a gentle fade out at the end to avoid any clicks
    fade_samples = int(0.1 * SAMPLE_RATE)
    if fade_samples < total_samples:
        fade_window = np.ones(total_samples)
        fade_window[-fade_samples:] = np.linspace(1.0, 0.0, fade_samples)
        
        output_signal[:, 0] *= fade_window
        output_signal[:, 1] *= fade_window
    
    # 8. Normalize to avoid clipping
    max_val = np.max(np.abs(output_signal))
    if max_val > 0:
        output_signal = output_signal / max_val * 0.9
    
    # Create distance array for plotting
    distances = np.sqrt(x_positions**2 + y_positions**2)
    
    return output_signal, distances

def plot_distance_trajectory(distances, azimuth, output_path):
    """Plot the trajectory of the approaching sound (distance over time)"""
    # Create figure
    plt.figure(figsize=(10, 6))
    
    # Define time axis (in seconds)
    time_axis = np.linspace(0, DURATION, len(distances))
    
    # Plot distance vs time
    plt.plot(time_axis, distances, 'b-', linewidth=3)
    plt.scatter(time_axis, distances, c=distances, cmap='viridis', s=80, alpha=0.7)
    
    # Mark start and end points
    plt.scatter(time_axis[0], distances[0], color='green', s=120, marker='o', label="Start (Far)")
    plt.scatter(time_axis[-1], distances[-1], color='red', s=120, marker='o', label="End (Near)")
    
    # Add grid and labels
    plt.grid(True, linestyle='--', alpha=0.5)
    
    # Create azimuth direction string
    if azimuth == 0:
        direction = "Front"
    elif azimuth == 90:
        direction = "Right"
    elif azimuth == -90:
        direction = "Left"
    else:
        direction = f"{azimuth}°"
    
    plt.title(f"Sound Source Approaching from the {direction} ({azimuth}° azimuth)", fontsize=14)
    plt.xlabel("Time (seconds)", fontsize=12)
    plt.ylabel("Distance from Listener (meters)", fontsize=12)
    plt.legend(fontsize=10)
    
    # Add annotations
    plt.annotate("Far", (time_axis[0], distances[0]), 
                xytext=(time_axis[0]-0.2, distances[0]+0.2),
                fontsize=12, fontweight='bold')
    plt.annotate("Near", (time_axis[-1], distances[-1]), 
                xytext=(time_axis[-1]-0.2, distances[-1]+0.2),
                fontsize=12, fontweight='bold')
    
    # Add top-down view of the scene
    ax_inset = plt.axes([0.65, 0.15, 0.3, 0.3])  # [left, bottom, width, height]
    
    # Plot listener (center) and sound positions
    ax_inset.scatter([0], [0], color='black', s=200, marker='o', label="Listener")
    
    # Calculate positions along azimuth at different distances
    # We rotate by 90° to convert from azimuth to Cartesian
    x_positions = distances * np.cos(np.deg2rad(azimuth + 90))
    y_positions = distances * np.sin(np.deg2rad(azimuth + 90))
    
    # Plot trajectory points
    ax_inset.scatter(x_positions, y_positions, c=np.arange(len(distances)), cmap='viridis', s=50, alpha=0.7)
    ax_inset.plot(x_positions, y_positions, 'k-', alpha=0.5)
    
    # Mark start and end
    ax_inset.scatter(x_positions[0], y_positions[0], color='green', s=80, marker='o')
    ax_inset.scatter(x_positions[-1], y_positions[-1], color='red', s=80, marker='o')
    
    # Add an arrow to show direction of movement
    arrow_idx = len(distances) // 2
    ax_inset.arrow(x_positions[arrow_idx], y_positions[arrow_idx], 
                  (x_positions[arrow_idx+1] - x_positions[arrow_idx])*3, 
                  (y_positions[arrow_idx+1] - y_positions[arrow_idx])*3,
                  head_width=0.1, head_length=0.15, fc='red', ec='red')
    
    # Draw head
    head_circle = plt.Circle((0, 0), HEAD_DIAMETER/2, fill=False, color='blue', alpha=0.7)
    ax_inset.add_artist(head_circle)
    
    # Draw ears
    ax_inset.scatter([-HEAD_DIAMETER/2, HEAD_DIAMETER/2], [0, 0], color='blue', s=50, marker='o')
    
    # Set equal aspect ratio for inset
    ax_inset.set_aspect('equal')
    ax_inset.set_title("Top-Down View", fontsize=10)
    ax_inset.grid(True, linestyle='--', alpha=0.3)
    ax_inset.set_xlabel("X (meters)")
    ax_inset.set_ylabel("Y (meters)")
    
    # Set axis limits to show the geometry
    max_range = max(abs(INITIAL_DISTANCE), abs(FINAL_DISTANCE)) * 1.2
    ax_inset.set_xlim(-max_range, max_range)
    ax_inset.set_ylim(-max_range, max_range)
    
    # Save figure
    plt.savefig(output_path, dpi=300, bbox_inches='tight')
    plt.close()
    
    print(f"Saved trajectory plot to: {output_path}")

def main():
    """Generate approaching sound from different directions using Fabian SOFA models with physics-based approach"""
    if not SOFA_AVAILABLE:
        print("ERROR: sofar library is required. Please install it with:")
        print("pip install sofar")
        return
    
    # Generate source signal (mono)
    print("Generating source signal...")
    source_signal = generate_pink_noise(DURATION, SAMPLE_RATE)
    
    # Process each SOFA file
    valid_files = []
    
    for sofa_path in sofa_files:
        if not os.path.exists(sofa_path):
            print(f"File not found: {sofa_path}")
            continue
        
        valid_files.append(sofa_path)
        file_name = os.path.basename(sofa_path).replace('.sofa', '')
        print(f"\nProcessing {file_name}...")
        
        try:
            # Load SOFA file
            hrtf, dimensions = load_sofa_file(sofa_path)
            
            if hrtf is None:
                print(f"Failed to load {file_name}, skipping.")
                continue
            
            # Create approaching movement for each azimuth
            for azimuth in AZIMUTH_ANGLES:
                print(f"\nGenerating approaching sound from azimuth {azimuth}°...")
                
                # Create approaching movement with physics-based model
                output_signal, distances = create_approaching_movement_physics(hrtf, source_signal, azimuth)
                
                # Save audio file
                if azimuth == 0:
                    direction = "front"
                elif azimuth == 90:
                    direction = "right"
                elif azimuth == -90:
                    direction = "left"
                else:
                    direction = f"az{azimuth}"
                
                output_filename = f"approaching_{direction}_az{azimuth}_{file_name}_physics.wav"
                output_path = os.path.join(output_dir, output_filename)
                sf.write(output_path, output_signal, SAMPLE_RATE)
                print(f"Saved audio to: {output_path}")
                
                # Create trajectory plot for each azimuth
                plot_path = os.path.join(output_dir, 
                                       f"trajectory_{direction}_az{azimuth}_{file_name}.png")
                plot_distance_trajectory(distances, azimuth, plot_path)
            
        except Exception as e:
            print(f"Error processing {file_name}: {e}")
            import traceback
            traceback.print_exc()
    
    print("\nApproaching movement processing complete!")
    print(f"Total SOFA models processed: {len(valid_files)}")
    print(f"All files saved to: {output_dir}")
    
    # Summarize settings used
    print("\nSettings used:")
    print(f"  Initial distance: {INITIAL_DISTANCE}m")
    print(f"  Final distance: {FINAL_DISTANCE}m")
    print(f"  Motion duration: {DURATION}s")
    print(f"  Azimuths: {AZIMUTH_ANGLES}")
    print(f"  Elevation: {CONSTANT_ELEVATION}°")
    print(f"  Head diameter: {HEAD_DIAMETER}m")
    print(f"  Distance attenuation factor: {DISTANCE_FACTOR}")

if __name__ == "__main__":
    main()
```

```python
"""
SOFA HRTF Moving Sound Source Simulator

This script creates binaural audio files simulating a sound source approaching the listener,
using the Fabian HRTF models for directional cues and a physically-based distance model
inspired by the EPFL LEMA approach.
"""
import numpy as np
import scipy.signal as signal
import soundfile as sf
import matplotlib.pyplot as plt
import os
import time
import datetime

# First, ensure required libraries are installed
try:
    import sofar
    SOFA_AVAILABLE = True
except ImportError:
    print("Warning: sofar library not found. Please install with: pip install sofar")
    SOFA_AVAILABLE = False

#############################################################
# CONTROL PARAMETERS - ADJUST THESE TO CHANGE THE SIMULATION
#############################################################

# Spatial parameters
INITIAL_DISTANCE = 2.0    # Starting distance in meters
FINAL_DISTANCE = 0.5      # Final distance in meters (increased to avoid extreme effects)
AZIMUTH_ANGLES = [90, 0, -90]  # Angles to generate (right, front, left)
CONSTANT_ELEVATION = 0    # Fixed elevation (ear level)
DISTANCE_FACTOR = 40      # Attenuation factor (higher = stronger distance effect)
HEAD_DIAMETER = 0.18      # Diameter of the head in meters

# Time parameters
DURATION = 3.0            # Duration of movement in seconds
SAMPLE_RATE = 44100       # Audio sample rate in Hz
SPEED_OF_SOUND = 343.0    # Speed of sound in m/s

# Use Fabian SOFA files - both HRIR and CTF files
sofa_files = [
    r"C:\Users\cogpsy-vrlab\Downloads\HRIRs_neutral_head_orientation_v4\SOFA\FABIAN_HRIR_measured_HATO_0.sofa",
    r"C:\Users\cogpsy-vrlab\Downloads\HRIRs_neutral_head_orientation_v4\SOFA\FABIAN_CTF_measured.sofa",
    r"C:\Users\cogpsy-vrlab\Downloads\HRIRs_neutral_head_orientation_v4\SOFA\FABIAN_CTF_measured_smoothed.sofa"
]

# Create output directory with timestamp
timestamp = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
output_dir = os.path.join(r"C:\Users\cogpsy-vrlab\Documents\PPS_module\BreathingPilot\LoomingStimuli", f"looming_stimuli_{timestamp}")
os.makedirs(output_dir, exist_ok=True)

def generate_pink_noise(duration, fs):
    """Generate pink noise using filter-based method (similar to MATLAB code)"""
    # Create filter coefficients for pink noise
    # These are similar to the coefficients used in the MATLAB code
    b = np.array([0.049922035, -0.095993537, 0.050612699, -0.004408786])
    a = np.array([1.0, -2.494956002, 2.017265875, -0.522189400])
    
    # Generate white noise
    np.random.seed(42)  # For reproducibility
    white_noise = np.random.randn(int(duration * fs))
    
    # Apply filter to get pink noise
    pink_noise = signal.lfilter(b, a, white_noise)
    
    # Normalize
    pink_noise = pink_noise / np.max(np.abs(pink_noise)) * 0.8
    
    return pink_noise

def load_sofa_file(sofa_path):
    """Load HRTF data from SOFA file"""
    try:
        print(f"Loading SOFA file: {os.path.basename(sofa_path)}")
        
        # Load the SOFA file
        hrtf = sofar.read_sofa(sofa_path)
        
        # Display basic SOFA info
        print(f"SOFA convention: {hrtf.GLOBAL_SOFAConventions} v{hrtf.GLOBAL_SOFAConventionsVersion}")
        
        # Get dimensions
        dimensions = {}
        for dim_name in ['M', 'R', 'N']:
            try:
                dimensions[dim_name] = hrtf.get_dimension(dim_name)
                print(f"{dim_name} = {dimensions[dim_name]}")
            except:
                dimensions[dim_name] = None
                print(f"{dim_name} = Unknown")
        
        return hrtf, dimensions
        
    except Exception as e:
        print(f"Error loading SOFA file: {e}")
        import traceback
        traceback.print_exc()
        return None, None

def find_nearest_hrtf(hrtf, target_azimuth, target_elevation=0, target_distance=1.0):
    """Find the measurement closest to the target azimuth, elevation, and distance"""
    try:
        # Check if this is a CTF file (Common Transfer Function)
        # CTF files typically have only one measurement/position
        is_ctf = False
        if "CTF" in hrtf.GLOBAL_SOFAConventions or "CTF" in str(hrtf):
            is_ctf = True
            
        if is_ctf or (hasattr(hrtf, 'Data_IR') and hrtf.Data_IR.shape[0] == 1 and hrtf.Data_IR.shape[1] == 1):
            print(f"Detected CTF file - using the single filter for azimuth {target_azimuth}°")
            # For CTF, there's only one measurement so return index 0
            # Try to get the position if available, otherwise use default
            if hasattr(hrtf, 'SourcePosition') and len(hrtf.SourcePosition) > 0:
                nearest_pos = hrtf.SourcePosition[0] if len(hrtf.SourcePosition.shape) == 1 else hrtf.SourcePosition[0,:]
            else:
                nearest_pos = np.array([0.0, 0.0, 0.0])  # Default position
            return 0, nearest_pos
    
        # Get source positions for regular HRTF files
        source_positions = np.array(hrtf.SourcePosition)
        
        # Handle different position array shapes
        if source_positions.shape[0] == 3 and len(source_positions.shape) == 2:
            source_positions = source_positions.T
        
        # Check if positions are in degrees or radians
        position_type = hrtf.SourcePosition_Type
        max_azimuth = np.max(np.abs(source_positions[:, 0]))
        positions_in_degrees = max_azimuth > 6.28
        
        if positions_in_degrees:
            # Both source positions and target are in degrees
            az_compare = target_azimuth
            el_compare = target_elevation
        else:
            # Source positions in radians, convert target to radians
            az_compare = np.deg2rad(target_azimuth)
            el_compare = np.deg2rad(target_elevation)
        
        # Handle negative azimuths if data is in 0-360 format
        if positions_in_degrees and target_azimuth < 0:
            # Check if we have negative azimuths in the data
            has_negative_az = np.any(source_positions[:, 0] < 0)
            if not has_negative_az:
                # Convert target from -180 to 180 format to 0 to 360 format
                az_compare = (target_azimuth + 360) % 360
        
        # Create distance metric focusing on azimuth and elevation
        if position_type == 'spherical':
            dists = np.sqrt(
                2 * (source_positions[:, 0] - az_compare)**2 + 
                (source_positions[:, 1] - el_compare)**2
            )
        else:
            # For Cartesian or unknown position type
            dists = np.abs(source_positions[:, 0] - az_compare)
                
        # Find nearest position
        nearest_idx = np.argmin(dists)
        nearest_pos = source_positions[nearest_idx]
        
        return nearest_idx, nearest_pos
        
    except Exception as e:
        print(f"Error finding nearest HRTF: {e}")
        import traceback
        traceback.print_exc()
        return 0, None

def extract_hrtf_data(hrtf, idx):
    """Extract HRTF impulse responses for a given index"""
    hrtf_data = hrtf.Data_IR
    
    # Check if this is a CTF file (Common Transfer Function)
    # CTF files often have shape (1, 1, N) or similar with only one filter
    is_ctf = False
    if "CTF" in hrtf.GLOBAL_SOFAConventions or "CTF" in str(hrtf):
        is_ctf = True
    
    # Handle CTF files (apply same filter to both ears)
    if is_ctf or hrtf_data.shape[0] == 1 and hrtf_data.shape[1] == 1:
        print("Detected CTF file - using same filter for both ears")
        try:
            # Get the single filter
            common_filter = hrtf_data[0, 0, :]
            return common_filter, common_filter
        except Exception as e:
            print(f"Error extracting CTF data: {e}")
            # Return simple impulse responses as fallback
            dummy_ir = np.zeros(256)
            dummy_ir[0] = 1.0
            return dummy_ir, dummy_ir
    
    # Regular HRTF files with separate left/right filters
    # Check dimensions and extract
    if hrtf_data.shape[0] == 2 and hrtf_data.shape[1] > 2:
        # Shape is (R, M, N)
        hrtf_left = hrtf_data[0, idx, :]
        hrtf_right = hrtf_data[1, idx, :]
    else:
        # Shape is (M, R, N)
        try:
            hrtf_left = hrtf_data[idx, 0, :]
            hrtf_right = hrtf_data[idx, 1, :]
        except:
            # Try other dimension orderings if the above fails
            try:
                if len(hrtf_data.shape) == 4:  # If 4D array
                    hrtf_left = hrtf_data[idx, 0, 0, :]
                    hrtf_right = hrtf_data[idx, 1, 0, :]
                else:
                    raise ValueError(f"Unexpected HRTF data shape: {hrtf_data.shape}")
            except Exception as e:
                print(f"Error extracting HRTF data: {e}")
                # Return empty IRs as fallback
                dummy_ir = np.zeros(256)
                dummy_ir[0] = 1.0
                return dummy_ir, dummy_ir
    
    return hrtf_left, hrtf_right

def create_approaching_movement_physics(hrtf, source_signal, azimuth, verbose=True):
    """
    Create audio where the sound approaches from a fixed azimuth using the EPFL LEMA
    physically-based approach for distance effects
    """
    # Calculate total samples
    total_samples = int(DURATION * SAMPLE_RATE)
    
    # Use HRTF for directional cues - find closest matching HRTF for our azimuth
    nearest_idx, nearest_pos = find_nearest_hrtf(hrtf, azimuth, CONSTANT_ELEVATION)
    hrtf_left, hrtf_right = extract_hrtf_data(hrtf, nearest_idx)
    
    if verbose:
        if nearest_pos is not None:
            try:
                # For spherical coordinates
                if hasattr(hrtf, 'SourcePosition_Type') and hrtf.SourcePosition_Type == 'spherical':
                    # Check if in degrees or radians
                    if isinstance(nearest_pos, np.ndarray) and len(nearest_pos) > 0:
                        if np.max(np.abs(nearest_pos[0])) > 6.28:
                            pos_str = f"({nearest_pos[0]:.1f}°, {nearest_pos[1]:.1f}°)"
                        else:
                            pos_str = f"({np.rad2deg(nearest_pos[0]):.1f}°, {np.rad2deg(nearest_pos[1]):.1f}°)"
                    else:
                        pos_str = str(nearest_pos)
                else:
                    # For Cartesian or other formats
                    if isinstance(nearest_pos, np.ndarray) and len(nearest_pos) >= 3:
                        pos_str = f"[{nearest_pos[0]:.2f}, {nearest_pos[1]:.2f}, {nearest_pos[2]:.2f}]"
                    else:
                        pos_str = str(nearest_pos)
            except:
                pos_str = str(nearest_pos)
        else:
            pos_str = "unknown"
            
        print(f"Using HRTF measurement at position {pos_str}")
    
    # Apply the HRTF to the entire source signal for directional cues
    print("Applying HRTF filter for directional cues...")
    dir_left = signal.fftconvolve(source_signal, hrtf_left, mode='same')
    dir_right = signal.fftconvolve(source_signal, hrtf_right, mode='same')
    
    # Now apply the physics-based approach from the MATLAB code for distance effects
    
    # 1. Create source coordinates as it approaches, similar to the MATLAB code
    # Set up the geometry (x, y) coordinates
    # Coordinates of the ears (left and right)
    coo_ear = np.array([[-HEAD_DIAMETER/2, 0], [HEAD_DIAMETER/2, 0]])
    
    # Calculate the y-coordinate based on azimuth
    # For straight approaches, y is constant, x changes
    if azimuth == 0:  # Coming from front
        x_start = -INITIAL_DISTANCE
        x_stop = -FINAL_DISTANCE
        y_value = 0
    elif azimuth == 90:  # Coming from right
        x_start = 0
        x_stop = 0
        y_value = INITIAL_DISTANCE
        y_stop = FINAL_DISTANCE
    elif azimuth == -90:  # Coming from left
        x_start = 0 
        x_stop = 0
        y_value = -INITIAL_DISTANCE
        y_stop = -FINAL_DISTANCE
    else:
        # For other angles, calculate the x and y components
        angle_rad = np.deg2rad(azimuth + 90)  # Convert to radians with 0 = front
        x_start = INITIAL_DISTANCE * np.cos(angle_rad)
        y_start = INITIAL_DISTANCE * np.sin(angle_rad)
        x_stop = FINAL_DISTANCE * np.cos(angle_rad)
        y_stop = FINAL_DISTANCE * np.sin(angle_rad)
    
    # Create source coordinates for each sample
    time_points = np.linspace(0, DURATION, total_samples)
    if azimuth == 90 or azimuth == -90:
        # For right or left approaches, x is constant, y changes
        x_positions = np.full(total_samples, x_start)
        if azimuth == 90:  # Right side
            y_positions = np.linspace(y_value, y_stop, total_samples)
        else:  # Left side
            y_positions = np.linspace(y_value, y_stop, total_samples)
    else:
        # For other angles
        x_positions = np.linspace(x_start, x_stop, total_samples)
        if azimuth == 0:  # Front
            y_positions = np.full(total_samples, y_value)
        else:
            y_positions = np.linspace(y_start, y_stop, total_samples)
    
    # Combine into source coordinates
    source_coordinates = np.column_stack((x_positions, y_positions))
    
    # 2. Calculate distances from the source to each ear at each time point
    # Distance from source to head center
    distance_to_head_center = np.sqrt(source_coordinates[:, 0]**2 + source_coordinates[:, 1]**2)
    
    # Calculate the direction of arrival angle for interaural time difference
    doa = np.zeros(total_samples)
    for i in range(total_samples):
        x_source = source_coordinates[i, 0]
        y_source = source_coordinates[i, 1]
        source_distance = np.sqrt(x_source**2 + y_source**2)
        
        if source_distance > 0:
            # Calculate angle in degrees
            angle_rad = np.arctan2(y_source, x_source)
            doa[i] = np.rad2deg(angle_rad)
        else:
            doa[i] = 0
    
    # Calculate distances from source to each ear
    # Using the spherical head model from the MATLAB code
    # Left ear
    dist_to_left_ear = np.zeros(total_samples)
    # Right ear
    dist_to_right_ear = np.zeros(total_samples)
    
    for i in range(total_samples):
        # Direct Euclidean distance from source to each ear
        x_source = source_coordinates[i, 0]
        y_source = source_coordinates[i, 1]
        
        # Distance to left ear
        dist_to_left_ear[i] = np.sqrt((x_source - coo_ear[0, 0])**2 + (y_source - coo_ear[0, 1])**2)
        
        # Distance to right ear
        dist_to_right_ear[i] = np.sqrt((x_source - coo_ear[1, 0])**2 + (y_source - coo_ear[1, 1])**2)
    
    # 3. Apply amplitude and delay based on distances
    # Calculate amplitude modulation using inverse square law with attenuation factor
    # This is adapted from the MATLAB code's approach
    ref_level = 100  # Reference sound level
    min_distance = 0.001  # To avoid division by zero
    
    # Use the reference amplitude and attenuation factor
    ref_amplitude = 2e-5 * 10**((ref_level-3)/20)
    
    # Calculate amplitude scaling based on distance
    left_amp = np.zeros(total_samples)
    right_amp = np.zeros(total_samples)
    
    for i in range(total_samples):
        # Amplitude based on distance, with directional bias
        d_left = max(dist_to_left_ear[i], min_distance)
        d_right = max(dist_to_right_ear[i], min_distance)
        d_center = max(distance_to_head_center[i], min_distance)
        
        # Calculate amplitude attenuation
        # Directional component based on angle (adapted from MATLAB)
        a_L = 5/180
        b_L = -5
        loss_dB_L = a_L * doa[i] + b_L
        
        a_R = -5/180
        b_R = 0
        loss_dB_R = a_R * doa[i] + b_R
        
        # Combine direction and distance effects
        left_amp[i] = 2e-5 * 10**((loss_dB_L + DISTANCE_FACTOR * np.log10(1/d_center))/20)
        right_amp[i] = 2e-5 * 10**((loss_dB_R + DISTANCE_FACTOR * np.log10(1/d_center))/20)
    
    # Normalize and scale
    left_amp = left_amp / np.max(left_amp) * ref_amplitude
    right_amp = right_amp / np.max(right_amp) * ref_amplitude
    
    # 4. Calculate time delays based on distance differences
    # Convert distance difference to time delay in samples
    left_delay_samples = np.round(dist_to_left_ear / SPEED_OF_SOUND * SAMPLE_RATE).astype(int)
    right_delay_samples = np.round(dist_to_right_ear / SPEED_OF_SOUND * SAMPLE_RATE).astype(int)
    
    # Find maximum delay to ensure we have enough buffer
    max_delay = max(np.max(left_delay_samples), np.max(right_delay_samples))
    
    # 5. Create output buffer with extra space for delays
    output_signal = np.zeros((total_samples + max_delay, 2))
    
    # 6. Apply the amplitude and delay modulation to the directional signal
    for i in range(total_samples):
        # Apply amplitude scaling and add to output with delay
        output_signal[left_delay_samples[i] + i, 0] += dir_left[i] * left_amp[i]
        output_signal[right_delay_samples[i] + i, 1] += dir_right[i] * right_amp[i]
    
    # Ensure we have enough samples and trim if needed
    if len(output_signal) > total_samples:
        output_signal = output_signal[:total_samples]
    
    # 7. Apply a gentle fade out at the end to avoid any clicks
    fade_samples = int(0.1 * SAMPLE_RATE)
    if fade_samples < total_samples:
        fade_window = np.ones(total_samples)
        fade_window[-fade_samples:] = np.linspace(1.0, 0.0, fade_samples)
        
        output_signal[:, 0] *= fade_window
        output_signal[:, 1] *= fade_window
    
    # 8. Normalize to avoid clipping
    max_val = np.max(np.abs(output_signal))
    if max_val > 0:
        output_signal = output_signal / max_val * 0.9
    
    # Create distance array for plotting
    distances = np.sqrt(x_positions**2 + y_positions**2)
    
    return output_signal, distances

def plot_distance_trajectory(distances, azimuth, output_path):
    """Plot the trajectory of the approaching sound (distance over time)"""
    # Create figure
    plt.figure(figsize=(10, 6))
    
    # Define time axis (in seconds)
    time_axis = np.linspace(0, DURATION, len(distances))
    
    # Plot distance vs time
    plt.plot(time_axis, distances, 'b-', linewidth=3)
    plt.scatter(time_axis, distances, c=distances, cmap='viridis', s=80, alpha=0.7)
    
    # Mark start and end points
    plt.scatter(time_axis[0], distances[0], color='green', s=120, marker='o', label="Start (Far)")
    plt.scatter(time_axis[-1], distances[-1], color='red', s=120, marker='o', label="End (Near)")
    
    # Add grid and labels
    plt.grid(True, linestyle='--', alpha=0.5)
    
    # Create azimuth direction string
    if azimuth == 0:
        direction = "Front"
    elif azimuth == 90:
        direction = "Right"
    elif azimuth == -90:
        direction = "Left"
    else:
        direction = f"{azimuth}°"
    
    plt.title(f"Sound Source Approaching from the {direction} ({azimuth}° azimuth)", fontsize=14)
    plt.xlabel("Time (seconds)", fontsize=12)
    plt.ylabel("Distance from Listener (meters)", fontsize=12)
    plt.legend(fontsize=10, loc='lower left')
    
    # Add annotations
    plt.annotate("Far", (time_axis[0], distances[0]), 
                xytext=(time_axis[0]-0.2, distances[0]+0.2),
                fontsize=12, fontweight='bold')
    plt.annotate("Near", (time_axis[-1], distances[-1]), 
                xytext=(time_axis[-1]-0.2, distances[-1]+0.2),
                fontsize=12, fontweight='bold')
    
    # Add top-down view of the scene
    ax_inset = plt.axes([0.65, 0.15, 0.3, 0.3])  # [left, bottom, width, height]
    
    # Plot listener (center) and sound positions
    ax_inset.scatter([0], [0], color='black', s=200, marker='o', label="Listener")
    
    # Calculate positions along azimuth at different distances
    # We rotate by 90° to convert from azimuth to Cartesian
    x_positions = distances * np.cos(np.deg2rad(azimuth + 90))
    y_positions = distances * np.sin(np.deg2rad(azimuth + 90))
    
    # Plot trajectory points
    ax_inset.scatter(x_positions, y_positions, c=np.arange(len(distances)), cmap='viridis', s=50, alpha=0.7)
    ax_inset.plot(x_positions, y_positions, 'k-', alpha=0.5)
    
    # Mark start and end
    ax_inset.scatter(x_positions[0], y_positions[0], color='green', s=80, marker='o')
    ax_inset.scatter(x_positions[-1], y_positions[-1], color='red', s=80, marker='o')
    
    # Add an arrow to show direction of movement
    arrow_idx = len(distances) // 2
    ax_inset.arrow(x_positions[arrow_idx], y_positions[arrow_idx], 
                  (x_positions[arrow_idx+1] - x_positions[arrow_idx])*3, 
                  (y_positions[arrow_idx+1] - y_positions[arrow_idx])*3,
                  head_width=0.1, head_length=0.15, fc='red', ec='red')
    
    # Draw head
    head_circle = plt.Circle((0, 0), HEAD_DIAMETER/2, fill=False, color='blue', alpha=0.7)
    ax_inset.add_artist(head_circle)
    
    # Draw ears
    ax_inset.scatter([-HEAD_DIAMETER/2, HEAD_DIAMETER/2], [0, 0], color='blue', s=50, marker='o')
    
    # Set equal aspect ratio for inset
    ax_inset.set_aspect('equal')
    ax_inset.set_title("Top-Down View", fontsize=10)
    ax_inset.grid(True, linestyle='--', alpha=0.3)
    ax_inset.set_xlabel("X (meters)")
    ax_inset.set_ylabel("Y (meters)")
    
    # Add legend to bottom left of inset plot
    ax_inset.legend(loc='lower left', fontsize=8)
    
    # Set axis limits to show the geometry
    max_range = max(abs(INITIAL_DISTANCE), abs(FINAL_DISTANCE)) * 1.2
    ax_inset.set_xlim(-max_range, max_range)
    ax_inset.set_ylim(-max_range, max_range)
    
    # Save figure
    plt.savefig(output_path, dpi=300, bbox_inches='tight')
    plt.close()
    
    print(f"Saved trajectory plot to: {output_path}")

def main():
    """Generate approaching sound from different directions using Fabian SOFA models with physics-based approach"""
    if not SOFA_AVAILABLE:
        print("ERROR: sofar library is required. Please install it with:")
        print("pip install sofar")
        return
    
    # Generate source signal (mono)
    print("Generating source signal...")
    source_signal = generate_pink_noise(DURATION, SAMPLE_RATE)
    
    # Process each SOFA file
    valid_files = []
    
    for sofa_path in sofa_files:
        if not os.path.exists(sofa_path):
            print(f"File not found: {sofa_path}")
            continue
        
        valid_files.append(sofa_path)
        file_name = os.path.basename(sofa_path).replace('.sofa', '')
        print(f"\nProcessing {file_name}...")
        
        try:
            # Load SOFA file
            hrtf, dimensions = load_sofa_file(sofa_path)
            
            if hrtf is None:
                print(f"Failed to load {file_name}, skipping.")
                continue
            
            # Create approaching movement for each azimuth
            for azimuth in AZIMUTH_ANGLES:
                print(f"\nGenerating approaching sound from azimuth {azimuth}°...")
                
                # Create approaching movement with physics-based model
                output_signal, distances = create_approaching_movement_physics(hrtf, source_signal, azimuth)
                
                # Save audio file
                if azimuth == 0:
                    direction = "front"
                elif azimuth == 90:
                    direction = "right"
                elif azimuth == -90:
                    direction = "left"
                else:
                    direction = f"az{azimuth}"
                
                output_filename = f"approaching_{direction}_az{azimuth}_{file_name}_physics.wav"
                output_path = os.path.join(output_dir, output_filename)
                sf.write(output_path, output_signal, SAMPLE_RATE)
                print(f"Saved audio to: {output_path}")
                
                # Create trajectory plot for each azimuth
                plot_path = os.path.join(output_dir, 
                                       f"trajectory_{direction}_az{azimuth}_{file_name}.png")
                plot_distance_trajectory(distances, azimuth, plot_path)
            
        except Exception as e:
            print(f"Error processing {file_name}: {e}")
            import traceback
            traceback.print_exc()
    
    print("\nApproaching movement processing complete!")
    print(f"Total SOFA models processed: {len(valid_files)}")
    print(f"All files saved to: {output_dir}")
    
    # Summarize settings used
    print("\nSettings used:")
    print(f"  Initial distance: {INITIAL_DISTANCE}m")
    print(f"  Final distance: {FINAL_DISTANCE}m")
    print(f"  Motion duration: {DURATION}s")
    print(f"  Azimuths: {AZIMUTH_ANGLES}")
    print(f"  Elevation: {CONSTANT_ELEVATION}°")
    print(f"  Head diameter: {HEAD_DIAMETER}m")
    print(f"  Distance attenuation factor: {DISTANCE_FACTOR}")

if __name__ == "__main__":
    main()
```

```python
"""
SOFA HRTF Moving Sound Source Simulator

This script creates binaural audio files simulating a sound source approaching the listener,
using the Fabian HRTF models for directional cues and a physically-based distance model
inspired by the EPFL LEMA approach.
"""
import numpy as np
import scipy.signal as signal
import soundfile as sf
import matplotlib.pyplot as plt
import os
import time
import datetime

# First, ensure required libraries are installed
try:
    import sofar
    SOFA_AVAILABLE = True
except ImportError:
    print("Warning: sofar library not found. Please install with: pip install sofar")
    SOFA_AVAILABLE = False

#############################################################
# CONTROL PARAMETERS - ADJUST THESE TO CHANGE THE SIMULATION
#############################################################

# Spatial parameters
INITIAL_DISTANCE = 2.0    # Starting distance in meters
FINAL_DISTANCE = 0.5      # Final distance in meters (increased to avoid extreme effects)
AZIMUTH_ANGLES = [90, 0, -90]  # Angles to generate (right, front, left)
CONSTANT_ELEVATION = 0    # Fixed elevation (ear level)
DISTANCE_FACTOR = 40      # Attenuation factor (higher = stronger distance effect)
HEAD_DIAMETER = 0.22      # Diameter of the head in meters (increased for stronger binaural effect)
BINAURAL_EMPHASIS = 1.5   # Factor to emphasize binaural cues (1.0 = normal)

# Time parameters
DURATION = 3.0            # Duration of movement in seconds
SAMPLE_RATE = 44100       # Audio sample rate in Hz
SPEED_OF_SOUND = 343.0    # Speed of sound in m/s

# Use only HRIR files for maximum binaural effect
sofa_files = [
    r"C:\Users\cogpsy-vrlab\Downloads\HRIRs_neutral_head_orientation_v4\SOFA\FABIAN_HRIR_measured_HATO_0.sofa"
]

# Create output directory with timestamp
timestamp = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
output_dir = os.path.join(r"C:\Users\cogpsy-vrlab\Documents\PPS_module\BreathingPilot\LoomingStimuli", f"looming_stimuli_{timestamp}")
os.makedirs(output_dir, exist_ok=True)

def generate_pink_noise(duration, fs):
    """Generate pink noise using filter-based method (similar to MATLAB code)"""
    # Create filter coefficients for pink noise
    # These are similar to the coefficients used in the MATLAB code
    b = np.array([0.049922035, -0.095993537, 0.050612699, -0.004408786])
    a = np.array([1.0, -2.494956002, 2.017265875, -0.522189400])
    
    # Generate white noise
    np.random.seed(42)  # For reproducibility
    white_noise = np.random.randn(int(duration * fs))
    
    # Apply filter to get pink noise
    pink_noise = signal.lfilter(b, a, white_noise)
    
    # Normalize
    pink_noise = pink_noise / np.max(np.abs(pink_noise)) * 0.8
    
    return pink_noise

def load_sofa_file(sofa_path):
    """Load HRTF data from SOFA file"""
    try:
        print(f"Loading SOFA file: {os.path.basename(sofa_path)}")
        
        # Load the SOFA file
        hrtf = sofar.read_sofa(sofa_path)
        
        # Display basic SOFA info
        print(f"SOFA convention: {hrtf.GLOBAL_SOFAConventions} v{hrtf.GLOBAL_SOFAConventionsVersion}")
        
        # Get dimensions
        dimensions = {}
        for dim_name in ['M', 'R', 'N']:
            try:
                dimensions[dim_name] = hrtf.get_dimension(dim_name)
                print(f"{dim_name} = {dimensions[dim_name]}")
            except:
                dimensions[dim_name] = None
                print(f"{dim_name} = Unknown")
        
        return hrtf, dimensions
        
    except Exception as e:
        print(f"Error loading SOFA file: {e}")
        import traceback
        traceback.print_exc()
        return None, None

def find_nearest_hrtf(hrtf, target_azimuth, target_elevation=0, target_distance=1.0):
    """Find the measurement closest to the target azimuth, elevation, and distance"""
    try:
        # Check if this is a CTF file (Common Transfer Function)
        # CTF files typically have only one measurement/position
        is_ctf = False
        if "CTF" in hrtf.GLOBAL_SOFAConventions or "CTF" in str(hrtf):
            is_ctf = True
            
        if is_ctf or (hasattr(hrtf, 'Data_IR') and hrtf.Data_IR.shape[0] == 1 and hrtf.Data_IR.shape[1] == 1):
            print(f"Detected CTF file - using the single filter for azimuth {target_azimuth}°")
            # For CTF, there's only one measurement so return index 0
            # Try to get the position if available, otherwise use default
            if hasattr(hrtf, 'SourcePosition') and len(hrtf.SourcePosition) > 0:
                nearest_pos = hrtf.SourcePosition[0] if len(hrtf.SourcePosition.shape) == 1 else hrtf.SourcePosition[0,:]
            else:
                nearest_pos = np.array([0.0, 0.0, 0.0])  # Default position
            return 0, nearest_pos
    
        # Get source positions for regular HRTF files
        source_positions = np.array(hrtf.SourcePosition)
        
        # Handle different position array shapes
        if source_positions.shape[0] == 3 and len(source_positions.shape) == 2:
            source_positions = source_positions.T
        
        # Check if positions are in degrees or radians
        position_type = hrtf.SourcePosition_Type
        max_azimuth = np.max(np.abs(source_positions[:, 0]))
        positions_in_degrees = max_azimuth > 6.28
        
        if positions_in_degrees:
            # Both source positions and target are in degrees
            az_compare = target_azimuth
            el_compare = target_elevation
        else:
            # Source positions in radians, convert target to radians
            az_compare = np.deg2rad(target_azimuth)
            el_compare = np.deg2rad(target_elevation)
        
        # Handle negative azimuths if data is in 0-360 format
        if positions_in_degrees and target_azimuth < 0:
            # Check if we have negative azimuths in the data
            has_negative_az = np.any(source_positions[:, 0] < 0)
            if not has_negative_az:
                # Convert target from -180 to 180 format to 0 to 360 format
                az_compare = (target_azimuth + 360) % 360
        
        # Create distance metric focusing on azimuth and elevation
        if position_type == 'spherical':
            dists = np.sqrt(
                2 * (source_positions[:, 0] - az_compare)**2 + 
                (source_positions[:, 1] - el_compare)**2
            )
        else:
            # For Cartesian or unknown position type
            dists = np.abs(source_positions[:, 0] - az_compare)
                
        # Find nearest position
        nearest_idx = np.argmin(dists)
        nearest_pos = source_positions[nearest_idx]
        
        return nearest_idx, nearest_pos
        
    except Exception as e:
        print(f"Error finding nearest HRTF: {e}")
        import traceback
        traceback.print_exc()
        return 0, None

def extract_hrtf_data(hrtf, idx):
    """Extract HRTF impulse responses for a given index"""
    hrtf_data = hrtf.Data_IR
    
    # Check if this is a CTF file (Common Transfer Function)
    # CTF files often have shape (1, 1, N) or similar with only one filter
    is_ctf = False
    if "CTF" in hrtf.GLOBAL_SOFAConventions or "CTF" in str(hrtf):
        is_ctf = True
    
    # Handle CTF files (apply same filter to both ears)
    if is_ctf or hrtf_data.shape[0] == 1 and hrtf_data.shape[1] == 1:
        print("Detected CTF file - using same filter for both ears")
        try:
            # Get the single filter
            common_filter = hrtf_data[0, 0, :]
            return common_filter, common_filter
        except Exception as e:
            print(f"Error extracting CTF data: {e}")
            # Return simple impulse responses as fallback
            dummy_ir = np.zeros(256)
            dummy_ir[0] = 1.0
            return dummy_ir, dummy_ir
    
    # Regular HRTF files with separate left/right filters
    # Check dimensions and extract
    if hrtf_data.shape[0] == 2 and hrtf_data.shape[1] > 2:
        # Shape is (R, M, N)
        hrtf_left = hrtf_data[0, idx, :]
        hrtf_right = hrtf_data[1, idx, :]
    else:
        # Shape is (M, R, N)
        try:
            hrtf_left = hrtf_data[idx, 0, :]
            hrtf_right = hrtf_data[idx, 1, :]
        except:
            # Try other dimension orderings if the above fails
            try:
                if len(hrtf_data.shape) == 4:  # If 4D array
                    hrtf_left = hrtf_data[idx, 0, 0, :]
                    hrtf_right = hrtf_data[idx, 1, 0, :]
                else:
                    raise ValueError(f"Unexpected HRTF data shape: {hrtf_data.shape}")
            except Exception as e:
                print(f"Error extracting HRTF data: {e}")
                # Return empty IRs as fallback
                dummy_ir = np.zeros(256)
                dummy_ir[0] = 1.0
                return dummy_ir, dummy_ir
    
    return hrtf_left, hrtf_right

def create_approaching_movement_physics(hrtf, source_signal, azimuth, verbose=True):
    """
    Create audio where the sound approaches from a fixed azimuth using the EPFL LEMA
    physically-based approach for distance effects
    """
    # Calculate total samples
    total_samples = int(DURATION * SAMPLE_RATE)
    
    # Use HRTF for directional cues - find closest matching HRTF for our azimuth
    nearest_idx, nearest_pos = find_nearest_hrtf(hrtf, azimuth, CONSTANT_ELEVATION)
    hrtf_left, hrtf_right = extract_hrtf_data(hrtf, nearest_idx)
    
    if verbose:
        if nearest_pos is not None:
            try:
                # For spherical coordinates
                if hasattr(hrtf, 'SourcePosition_Type') and hrtf.SourcePosition_Type == 'spherical':
                    # Check if in degrees or radians
                    if isinstance(nearest_pos, np.ndarray) and len(nearest_pos) > 0:
                        if np.max(np.abs(nearest_pos[0])) > 6.28:
                            pos_str = f"({nearest_pos[0]:.1f}°, {nearest_pos[1]:.1f}°)"
                        else:
                            pos_str = f"({np.rad2deg(nearest_pos[0]):.1f}°, {np.rad2deg(nearest_pos[1]):.1f}°)"
                    else:
                        pos_str = str(nearest_pos)
                else:
                    # For Cartesian or other formats
                    if isinstance(nearest_pos, np.ndarray) and len(nearest_pos) >= 3:
                        pos_str = f"[{nearest_pos[0]:.2f}, {nearest_pos[1]:.2f}, {nearest_pos[2]:.2f}]"
                    else:
                        pos_str = str(nearest_pos)
            except:
                pos_str = str(nearest_pos)
        else:
            pos_str = "unknown"
            
        print(f"Using HRTF measurement at position {pos_str}")
    
    # Apply the HRTF to the entire source signal for directional cues
    print("Applying HRTF filter for directional cues...")
    dir_left = signal.fftconvolve(source_signal, hrtf_left, mode='same')
    dir_right = signal.fftconvolve(source_signal, hrtf_right, mode='same')
    
    # Now apply the physics-based approach from the MATLAB code for distance effects
    
    # 1. Create source coordinates as it approaches, similar to the MATLAB code
    # Set up the geometry (x, y) coordinates
    # Coordinates of the ears (left and right)
    coo_ear = np.array([[-HEAD_DIAMETER/2, 0], [HEAD_DIAMETER/2, 0]])
    
    # Calculate the y-coordinate based on azimuth
    # For straight approaches, y is constant, x changes
    if azimuth == 0:  # Coming from front
        x_start = -INITIAL_DISTANCE
        x_stop = -FINAL_DISTANCE
        y_value = 0
    elif azimuth == 90:  # Coming from right
        x_start = 0
        x_stop = 0
        y_value = INITIAL_DISTANCE
        y_stop = FINAL_DISTANCE
    elif azimuth == -90:  # Coming from left
        x_start = 0 
        x_stop = 0
        y_value = -INITIAL_DISTANCE
        y_stop = -FINAL_DISTANCE
    else:
        # For other angles, calculate the x and y components
        angle_rad = np.deg2rad(azimuth + 90)  # Convert to radians with 0 = front
        x_start = INITIAL_DISTANCE * np.cos(angle_rad)
        y_start = INITIAL_DISTANCE * np.sin(angle_rad)
        x_stop = FINAL_DISTANCE * np.cos(angle_rad)
        y_stop = FINAL_DISTANCE * np.sin(angle_rad)
    
    # Create source coordinates for each sample
    time_points = np.linspace(0, DURATION, total_samples)
    if azimuth == 90 or azimuth == -90:
        # For right or left approaches, x is constant, y changes
        x_positions = np.full(total_samples, x_start)
        if azimuth == 90:  # Right side
            y_positions = np.linspace(y_value, y_stop, total_samples)
        else:  # Left side
            y_positions = np.linspace(y_value, y_stop, total_samples)
    else:
        # For other angles
        x_positions = np.linspace(x_start, x_stop, total_samples)
        if azimuth == 0:  # Front
            y_positions = np.full(total_samples, y_value)
        else:
            y_positions = np.linspace(y_start, y_stop, total_samples)
    
    # Combine into source coordinates
    source_coordinates = np.column_stack((x_positions, y_positions))
    
    # 2. Calculate distances from the source to each ear at each time point
    # Distance from source to head center
    distance_to_head_center = np.sqrt(source_coordinates[:, 0]**2 + source_coordinates[:, 1]**2)
    
    # Calculate the direction of arrival angle for interaural time difference
    doa = np.zeros(total_samples)
    for i in range(total_samples):
        x_source = source_coordinates[i, 0]
        y_source = source_coordinates[i, 1]
        source_distance = np.sqrt(x_source**2 + y_source**2)
        
        if source_distance > 0:
            # Calculate angle in degrees
            angle_rad = np.arctan2(y_source, x_source)
            doa[i] = np.rad2deg(angle_rad)
        else:
            doa[i] = 0
    
    # Calculate distances from source to each ear
    # Using the spherical head model from the MATLAB code
    # Left ear
    dist_to_left_ear = np.zeros(total_samples)
    # Right ear
    dist_to_right_ear = np.zeros(total_samples)
    
    for i in range(total_samples):
        # Direct Euclidean distance from source to each ear
        x_source = source_coordinates[i, 0]
        y_source = source_coordinates[i, 1]
        
        # Distance to left ear
        dist_to_left_ear[i] = np.sqrt((x_source - coo_ear[0, 0])**2 + (y_source - coo_ear[0, 1])**2)
        
        # Distance to right ear
        dist_to_right_ear[i] = np.sqrt((x_source - coo_ear[1, 0])**2 + (y_source - coo_ear[1, 1])**2)
    
    # 3. Apply amplitude and delay based on distances
    # Calculate amplitude modulation using inverse square law with attenuation factor
    # This is adapted from the MATLAB code's approach
    ref_level = 100  # Reference sound level
    min_distance = 0.001  # To avoid division by zero
    
    # Use the reference amplitude and attenuation factor
    ref_amplitude = 2e-5 * 10**((ref_level-3)/20)
    
    # Calculate amplitude scaling based on distance
    left_amp = np.zeros(total_samples)
    right_amp = np.zeros(total_samples)
    
    for i in range(total_samples):
        # Amplitude based on distance, with directional bias
        d_left = max(dist_to_left_ear[i], min_distance)
        d_right = max(dist_to_right_ear[i], min_distance)
        d_center = max(distance_to_head_center[i], min_distance)
        
        # Calculate amplitude attenuation
        # Directional component based on angle (adapted from MATLAB)
        a_L = 5/180
        b_L = -5
        loss_dB_L = a_L * doa[i] + b_L
        
        a_R = -5/180
        b_R = 0
        loss_dB_R = a_R * doa[i] + b_R
        
        # Combine direction and distance effects
        left_amp[i] = 2e-5 * 10**((loss_dB_L + DISTANCE_FACTOR * np.log10(1/d_center))/20)
        right_amp[i] = 2e-5 * 10**((loss_dB_R + DISTANCE_FACTOR * np.log10(1/d_center))/20)
    
    # Normalize and scale
    left_amp = left_amp / np.max(left_amp) * ref_amplitude
    right_amp = right_amp / np.max(right_amp) * ref_amplitude
    
    # 4. Calculate time delays based on distance differences
    # Convert distance difference to time delay in samples with enhanced ITD
    # Multiply by binaural emphasis factor to enhance interaural time differences
    left_delay_samples = np.round(BINAURAL_EMPHASIS * dist_to_left_ear / SPEED_OF_SOUND * SAMPLE_RATE).astype(int)
    right_delay_samples = np.round(BINAURAL_EMPHASIS * dist_to_right_ear / SPEED_OF_SOUND * SAMPLE_RATE).astype(int)
    
    # Find maximum delay to ensure we have enough buffer
    max_delay = max(np.max(left_delay_samples), np.max(right_delay_samples))
    
    # 5. Create output buffer with extra space for delays
    output_signal = np.zeros((total_samples + max_delay, 2))
    
    # 6. Apply the amplitude and delay modulation to the directional signal
    for i in range(total_samples):
        # Apply amplitude scaling and add to output with delay
        output_signal[left_delay_samples[i] + i, 0] += dir_left[i] * left_amp[i]
        output_signal[right_delay_samples[i] + i, 1] += dir_right[i] * right_amp[i]
    
    # Ensure we have enough samples and trim if needed
    if len(output_signal) > total_samples:
        output_signal = output_signal[:total_samples]
    
    # 7. Apply a gentle fade out at the end to avoid any clicks
    fade_samples = int(0.1 * SAMPLE_RATE)
    if fade_samples < total_samples:
        fade_window = np.ones(total_samples)
        fade_window[-fade_samples:] = np.linspace(1.0, 0.0, fade_samples)
        
        output_signal[:, 0] *= fade_window
        output_signal[:, 1] *= fade_window
    
    # 8. Normalize to avoid clipping
    max_val = np.max(np.abs(output_signal))
    if max_val > 0:
        output_signal = output_signal / max_val * 0.9
    
    # Create distance array for plotting
    distances = np.sqrt(x_positions**2 + y_positions**2)
    
    return output_signal, distances

def plot_distance_trajectory(distances, azimuth, output_path):
    """Plot the trajectory of the approaching sound (distance over time)"""
    # Create figure
    plt.figure(figsize=(10, 6))
    
    # Define time axis (in seconds)
    time_axis = np.linspace(0, DURATION, len(distances))
    
    # Plot distance vs time
    plt.plot(time_axis, distances, 'b-', linewidth=3)
    plt.scatter(time_axis, distances, c=distances, cmap='viridis', s=80, alpha=0.7)
    
    # Mark start and end points
    plt.scatter(time_axis[0], distances[0], color='green', s=120, marker='o', label="Start (Far)")
    plt.scatter(time_axis[-1], distances[-1], color='red', s=120, marker='o', label="End (Near)")
    
    # Add grid and labels
    plt.grid(True, linestyle='--', alpha=0.5)
    
    # Create azimuth direction string
    if azimuth == 0:
        direction = "Front"
    elif azimuth == 90:
        direction = "Right"
    elif azimuth == -90:
        direction = "Left"
    else:
        direction = f"{azimuth}°"
    
    plt.title(f"Sound Source Approaching from the {direction} ({azimuth}° azimuth)", fontsize=14)
    plt.xlabel("Time (seconds)", fontsize=12)
    plt.ylabel("Distance from Listener (meters)", fontsize=12)
    plt.legend(fontsize=10, loc='lower left')
    
    # Add annotations
    plt.annotate("Far", (time_axis[0], distances[0]), 
                xytext=(time_axis[0]-0.2, distances[0]+0.2),
                fontsize=12, fontweight='bold')
    plt.annotate("Near", (time_axis[-1], distances[-1]), 
                xytext=(time_axis[-1]-0.2, distances[-1]+0.2),
                fontsize=12, fontweight='bold')
    
    # Add top-down view of the scene
    ax_inset = plt.axes([0.65, 0.15, 0.3, 0.3])  # [left, bottom, width, height]
    
    # Plot listener (center) and sound positions
    ax_inset.scatter([0], [0], color='black', s=200, marker='o', label="Listener")
    
    # Calculate positions along azimuth at different distances
    # We rotate by 90° to convert from azimuth to Cartesian
    x_positions = distances * np.cos(np.deg2rad(azimuth + 90))
    y_positions = distances * np.sin(np.deg2rad(azimuth + 90))
    
    # Plot trajectory points
    ax_inset.scatter(x_positions, y_positions, c=np.arange(len(distances)), cmap='viridis', s=50, alpha=0.7)
    ax_inset.plot(x_positions, y_positions, 'k-', alpha=0.5)
    
    # Mark start and end
    ax_inset.scatter(x_positions[0], y_positions[0], color='green', s=80, marker='o')
    ax_inset.scatter(x_positions[-1], y_positions[-1], color='red', s=80, marker='o')
    
    # Add an arrow to show direction of movement
    arrow_idx = len(distances) // 2
    ax_inset.arrow(x_positions[arrow_idx], y_positions[arrow_idx], 
                  (x_positions[arrow_idx+1] - x_positions[arrow_idx])*3, 
                  (y_positions[arrow_idx+1] - y_positions[arrow_idx])*3,
                  head_width=0.1, head_length=0.15, fc='red', ec='red')
    
    # Draw head
    head_circle = plt.Circle((0, 0), HEAD_DIAMETER/2, fill=False, color='blue', alpha=0.7)
    ax_inset.add_artist(head_circle)
    
    # Draw ears
    ax_inset.scatter([-HEAD_DIAMETER/2, HEAD_DIAMETER/2], [0, 0], color='blue', s=50, marker='o')
    
    # Set equal aspect ratio for inset
    ax_inset.set_aspect('equal')
    ax_inset.set_title("Top-Down View", fontsize=10)
    ax_inset.grid(True, linestyle='--', alpha=0.3)
    ax_inset.set_xlabel("X (meters)")
    ax_inset.set_ylabel("Y (meters)")
    
    # Add legend to bottom left of inset plot
    ax_inset.legend(loc='lower left', fontsize=8)
    
    # Set axis limits to show the geometry
    max_range = max(abs(INITIAL_DISTANCE), abs(FINAL_DISTANCE)) * 1.2
    ax_inset.set_xlim(-max_range, max_range)
    ax_inset.set_ylim(-max_range, max_range)
    
    # Save figure
    plt.savefig(output_path, dpi=300, bbox_inches='tight')
    plt.close()
    
    print(f"Saved trajectory plot to: {output_path}")

def main():
    """Generate approaching sound from different directions using Fabian SOFA models with physics-based approach"""
    if not SOFA_AVAILABLE:
        print("ERROR: sofar library is required. Please install it with:")
        print("pip install sofar")
        return
    
    # Generate source signal (mono)
    print("Generating source signal...")
    source_signal = generate_pink_noise(DURATION, SAMPLE_RATE)
    
    # Process each SOFA file
    valid_files = []
    
    for sofa_path in sofa_files:
        if not os.path.exists(sofa_path):
            print(f"File not found: {sofa_path}")
            continue
        
        valid_files.append(sofa_path)
        file_name = os.path.basename(sofa_path).replace('.sofa', '')
        print(f"\nProcessing {file_name}...")
        
        try:
            # Load SOFA file
            hrtf, dimensions = load_sofa_file(sofa_path)
            
            if hrtf is None:
                print(f"Failed to load {file_name}, skipping.")
                continue
            
            # Create approaching movement for each azimuth
            for azimuth in AZIMUTH_ANGLES:
                print(f"\nGenerating approaching sound from azimuth {azimuth}°...")
                
                # Create approaching movement with physics-based model
                output_signal, distances = create_approaching_movement_physics(hrtf, source_signal, azimuth)
                
                # Save audio file
                if azimuth == 0:
                    direction = "front"
                elif azimuth == 90:
                    direction = "right"
                elif azimuth == -90:
                    direction = "left"
                else:
                    direction = f"az{azimuth}"
                
                output_filename = f"approaching_{direction}_az{azimuth}_FABIAN_HRIR_enhanced_binaural.wav"
                output_path = os.path.join(output_dir, output_filename)
                sf.write(output_path, output_signal, SAMPLE_RATE)
                print(f"Saved audio to: {output_path}")
                
                # Create trajectory plot for each azimuth
                plot_path = os.path.join(output_dir, 
                                       f"trajectory_{direction}_az{azimuth}_{file_name}.png")
                plot_distance_trajectory(distances, azimuth, plot_path)
            
        except Exception as e:
            print(f"Error processing {file_name}: {e}")
            import traceback
            traceback.print_exc()
    
    print("\nApproaching movement processing complete!")
    print(f"Total SOFA models processed: {len(valid_files)}")
    print(f"All files saved to: {output_dir}")
    
    # Summarize settings used
    print("\nSettings used:")
    print(f"  Initial distance: {INITIAL_DISTANCE}m")
    print(f"  Final distance: {FINAL_DISTANCE}m")
    print(f"  Motion duration: {DURATION}s")
    print(f"  Azimuths: {AZIMUTH_ANGLES}")
    print(f"  Elevation: {CONSTANT_ELEVATION}°")
    print(f"  Head diameter: {HEAD_DIAMETER}m (enhanced for stronger binaural effect)")
    print(f"  Distance attenuation factor: {DISTANCE_FACTOR}")
    print(f"  Binaural emphasis factor: {BINAURAL_EMPHASIS}")
    print(f"  HRTF model used: FABIAN_HRIR_measured_HATO_0 (for maximum binaural effect)")

if __name__ == "__main__":
    main()
```

```python
"""
SOFA HRTF Moving Sound Source Simulator

This script creates binaural audio files simulating a sound source approaching the listener,
using the Fabian HRTF models for directional cues and a physically-based distance model
inspired by the EPFL LEMA approach.
"""
import numpy as np
import scipy.signal as signal
import soundfile as sf
import matplotlib.pyplot as plt
import os
import time
import datetime

# First, ensure required libraries are installed
try:
    import sofar
    SOFA_AVAILABLE = True
except ImportError:
    print("Warning: sofar library not found. Please install with: pip install sofar")
    SOFA_AVAILABLE = False

#############################################################
# CONTROL PARAMETERS - ADJUST THESE TO CHANGE THE SIMULATION
#############################################################

# Spatial parameters
INITIAL_DISTANCE = 2.0    # Starting distance in meters
FINAL_DISTANCE = 0.2      # Final distance in meters (increased to avoid extreme effects)
AZIMUTH_ANGLES = [90, 0, -90]  # Angles to generate (right, front, left)
CONSTANT_ELEVATION = 0    # Fixed elevation (ear level)
DISTANCE_FACTOR = 40      # Attenuation factor (higher = stronger distance effect)
HEAD_DIAMETER = 0.18      # Diameter of the head in meters (natural human head size)

# Time parameters
DURATION = 3.0            # Duration of movement in seconds
SAMPLE_RATE = 48000       # Audio sample rate in Hz
SPEED_OF_SOUND = 343.0    # Speed of sound in m/s

# Use only HRIR files for maximum binaural effect
sofa_files = [
    r"C:\Users\cogpsy-vrlab\Downloads\HRIRs_neutral_head_orientation_v4\SOFA\FABIAN_HRIR_measured_HATO_0.sofa"
]

# Create output directory with timestamp
timestamp = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
output_dir = os.path.join(r"C:\Users\cogpsy-vrlab\Documents\PPS_module\BreathingPilot\LoomingStimuli", f"looming_stimuli_{timestamp}")
os.makedirs(output_dir, exist_ok=True)

def generate_pink_noise(duration, fs):
    """Generate pink noise using filter-based method (similar to MATLAB code)"""
    # Create filter coefficients for pink noise
    # These are similar to the coefficients used in the MATLAB code
    b = np.array([0.049922035, -0.095993537, 0.050612699, -0.004408786])
    a = np.array([1.0, -2.494956002, 2.017265875, -0.522189400])
    
    # Generate white noise
    np.random.seed(42)  # For reproducibility
    white_noise = np.random.randn(int(duration * fs))
    
    # Apply filter to get pink noise
    pink_noise = signal.lfilter(b, a, white_noise)
    
    # Normalize
    pink_noise = pink_noise / np.max(np.abs(pink_noise)) * 0.8
    
    return pink_noise

def load_sofa_file(sofa_path):
    """Load HRTF data from SOFA file"""
    try:
        print(f"Loading SOFA file: {os.path.basename(sofa_path)}")
        
        # Load the SOFA file
        hrtf = sofar.read_sofa(sofa_path)
        
        # Display basic SOFA info
        print(f"SOFA convention: {hrtf.GLOBAL_SOFAConventions} v{hrtf.GLOBAL_SOFAConventionsVersion}")
        
        # Get dimensions
        dimensions = {}
        for dim_name in ['M', 'R', 'N']:
            try:
                dimensions[dim_name] = hrtf.get_dimension(dim_name)
                print(f"{dim_name} = {dimensions[dim_name]}")
            except:
                dimensions[dim_name] = None
                print(f"{dim_name} = Unknown")
        
        return hrtf, dimensions
        
    except Exception as e:
        print(f"Error loading SOFA file: {e}")
        import traceback
        traceback.print_exc()
        return None, None

def find_nearest_hrtf(hrtf, target_azimuth, target_elevation=0, target_distance=1.0):
    """Find the measurement closest to the target azimuth, elevation, and distance"""
    try:
        # Check if this is a CTF file (Common Transfer Function)
        # CTF files typically have only one measurement/position
        is_ctf = False
        if "CTF" in hrtf.GLOBAL_SOFAConventions or "CTF" in str(hrtf):
            is_ctf = True
            
        if is_ctf or (hasattr(hrtf, 'Data_IR') and hrtf.Data_IR.shape[0] == 1 and hrtf.Data_IR.shape[1] == 1):
            print(f"Detected CTF file - using the single filter for azimuth {target_azimuth}°")
            # For CTF, there's only one measurement so return index 0
            # Try to get the position if available, otherwise use default
            if hasattr(hrtf, 'SourcePosition') and len(hrtf.SourcePosition) > 0:
                nearest_pos = hrtf.SourcePosition[0] if len(hrtf.SourcePosition.shape) == 1 else hrtf.SourcePosition[0,:]
            else:
                nearest_pos = np.array([0.0, 0.0, 0.0])  # Default position
            return 0, nearest_pos
    
        # Get source positions for regular HRTF files
        source_positions = np.array(hrtf.SourcePosition)
        
        # Handle different position array shapes
        if source_positions.shape[0] == 3 and len(source_positions.shape) == 2:
            source_positions = source_positions.T
        
        # Check if positions are in degrees or radians
        position_type = hrtf.SourcePosition_Type
        max_azimuth = np.max(np.abs(source_positions[:, 0]))
        positions_in_degrees = max_azimuth > 6.28
        
        if positions_in_degrees:
            # Both source positions and target are in degrees
            az_compare = target_azimuth
            el_compare = target_elevation
        else:
            # Source positions in radians, convert target to radians
            az_compare = np.deg2rad(target_azimuth)
            el_compare = np.deg2rad(target_elevation)
        
        # Handle negative azimuths if data is in 0-360 format
        if positions_in_degrees and target_azimuth < 0:
            # Check if we have negative azimuths in the data
            has_negative_az = np.any(source_positions[:, 0] < 0)
            if not has_negative_az:
                # Convert target from -180 to 180 format to 0 to 360 format
                az_compare = (target_azimuth + 360) % 360
        
        # Create distance metric focusing on azimuth and elevation
        if position_type == 'spherical':
            dists = np.sqrt(
                2 * (source_positions[:, 0] - az_compare)**2 + 
                (source_positions[:, 1] - el_compare)**2
            )
        else:
            # For Cartesian or unknown position type
            dists = np.abs(source_positions[:, 0] - az_compare)
                
        # Find nearest position
        nearest_idx = np.argmin(dists)
        nearest_pos = source_positions[nearest_idx]
        
        return nearest_idx, nearest_pos
        
    except Exception as e:
        print(f"Error finding nearest HRTF: {e}")
        import traceback
        traceback.print_exc()
        return 0, None

def extract_hrtf_data(hrtf, idx):
    """Extract HRTF impulse responses for a given index"""
    hrtf_data = hrtf.Data_IR
    
    # Check if this is a CTF file (Common Transfer Function)
    # CTF files often have shape (1, 1, N) or similar with only one filter
    is_ctf = False
    if "CTF" in hrtf.GLOBAL_SOFAConventions or "CTF" in str(hrtf):
        is_ctf = True
    
    # Handle CTF files (apply same filter to both ears)
    if is_ctf or hrtf_data.shape[0] == 1 and hrtf_data.shape[1] == 1:
        print("Detected CTF file - using same filter for both ears")
        try:
            # Get the single filter
            common_filter = hrtf_data[0, 0, :]
            return common_filter, common_filter
        except Exception as e:
            print(f"Error extracting CTF data: {e}")
            # Return simple impulse responses as fallback
            dummy_ir = np.zeros(256)
            dummy_ir[0] = 1.0
            return dummy_ir, dummy_ir
    
    # Regular HRTF files with separate left/right filters
    # Check dimensions and extract
    if hrtf_data.shape[0] == 2 and hrtf_data.shape[1] > 2:
        # Shape is (R, M, N)
        hrtf_left = hrtf_data[0, idx, :]
        hrtf_right = hrtf_data[1, idx, :]
    else:
        # Shape is (M, R, N)
        try:
            hrtf_left = hrtf_data[idx, 0, :]
            hrtf_right = hrtf_data[idx, 1, :]
        except:
            # Try other dimension orderings if the above fails
            try:
                if len(hrtf_data.shape) == 4:  # If 4D array
                    hrtf_left = hrtf_data[idx, 0, 0, :]
                    hrtf_right = hrtf_data[idx, 1, 0, :]
                else:
                    raise ValueError(f"Unexpected HRTF data shape: {hrtf_data.shape}")
            except Exception as e:
                print(f"Error extracting HRTF data: {e}")
                # Return empty IRs as fallback
                dummy_ir = np.zeros(256)
                dummy_ir[0] = 1.0
                return dummy_ir, dummy_ir
    
    return hrtf_left, hrtf_right

def create_approaching_movement_physics(hrtf, source_signal, azimuth, verbose=True):
    """
    Create audio where the sound approaches from a fixed azimuth using the EPFL LEMA
    physically-based approach for distance effects
    """
    # Calculate total samples
    total_samples = int(DURATION * SAMPLE_RATE)
    
    # Use HRTF for directional cues - find closest matching HRTF for our azimuth
    nearest_idx, nearest_pos = find_nearest_hrtf(hrtf, azimuth, CONSTANT_ELEVATION)
    hrtf_left, hrtf_right = extract_hrtf_data(hrtf, nearest_idx)
    
    if verbose:
        if nearest_pos is not None:
            try:
                # For spherical coordinates
                if hasattr(hrtf, 'SourcePosition_Type') and hrtf.SourcePosition_Type == 'spherical':
                    # Check if in degrees or radians
                    if isinstance(nearest_pos, np.ndarray) and len(nearest_pos) > 0:
                        if np.max(np.abs(nearest_pos[0])) > 6.28:
                            pos_str = f"({nearest_pos[0]:.1f}°, {nearest_pos[1]:.1f}°)"
                        else:
                            pos_str = f"({np.rad2deg(nearest_pos[0]):.1f}°, {np.rad2deg(nearest_pos[1]):.1f}°)"
                    else:
                        pos_str = str(nearest_pos)
                else:
                    # For Cartesian or other formats
                    if isinstance(nearest_pos, np.ndarray) and len(nearest_pos) >= 3:
                        pos_str = f"[{nearest_pos[0]:.2f}, {nearest_pos[1]:.2f}, {nearest_pos[2]:.2f}]"
                    else:
                        pos_str = str(nearest_pos)
            except:
                pos_str = str(nearest_pos)
        else:
            pos_str = "unknown"
            
        print(f"Using HRTF measurement at position {pos_str}")
    
    # Apply the HRTF to the entire source signal for directional cues
    print("Applying HRTF filter for directional cues...")
    dir_left = signal.fftconvolve(source_signal, hrtf_left, mode='same')
    dir_right = signal.fftconvolve(source_signal, hrtf_right, mode='same')
    
    # Now apply the physics-based approach from the MATLAB code for distance effects
    
    # 1. Create source coordinates as it approaches, similar to the MATLAB code
    # Set up the geometry (x, y) coordinates
    # Coordinates of the ears (left and right)
    coo_ear = np.array([[-HEAD_DIAMETER/2, 0], [HEAD_DIAMETER/2, 0]])
    
    # Calculate the y-coordinate based on azimuth
    # For straight approaches, y is constant, x changes
    if azimuth == 0:  # Coming from front
        x_start = -INITIAL_DISTANCE
        x_stop = -FINAL_DISTANCE
        y_value = 0
    elif azimuth == 90:  # Coming from right
        x_start = 0
        x_stop = 0
        y_value = INITIAL_DISTANCE
        y_stop = FINAL_DISTANCE
    elif azimuth == -90:  # Coming from left
        x_start = 0 
        x_stop = 0
        y_value = -INITIAL_DISTANCE
        y_stop = -FINAL_DISTANCE
    else:
        # For other angles, calculate the x and y components
        angle_rad = np.deg2rad(azimuth + 90)  # Convert to radians with 0 = front
        x_start = INITIAL_DISTANCE * np.cos(angle_rad)
        y_start = INITIAL_DISTANCE * np.sin(angle_rad)
        x_stop = FINAL_DISTANCE * np.cos(angle_rad)
        y_stop = FINAL_DISTANCE * np.sin(angle_rad)
    
    # Create source coordinates for each sample
    time_points = np.linspace(0, DURATION, total_samples)
    if azimuth == 90 or azimuth == -90:
        # For right or left approaches, x is constant, y changes
        x_positions = np.full(total_samples, x_start)
        if azimuth == 90:  # Right side
            y_positions = np.linspace(y_value, y_stop, total_samples)
        else:  # Left side
            y_positions = np.linspace(y_value, y_stop, total_samples)
    else:
        # For other angles
        x_positions = np.linspace(x_start, x_stop, total_samples)
        if azimuth == 0:  # Front
            y_positions = np.full(total_samples, y_value)
        else:
            y_positions = np.linspace(y_start, y_stop, total_samples)
    
    # Combine into source coordinates
    source_coordinates = np.column_stack((x_positions, y_positions))
    
    # 2. Calculate distances from the source to each ear at each time point
    # Distance from source to head center
    distance_to_head_center = np.sqrt(source_coordinates[:, 0]**2 + source_coordinates[:, 1]**2)
    
    # Calculate the direction of arrival angle for interaural time difference
    doa = np.zeros(total_samples)
    for i in range(total_samples):
        x_source = source_coordinates[i, 0]
        y_source = source_coordinates[i, 1]
        source_distance = np.sqrt(x_source**2 + y_source**2)
        
        if source_distance > 0:
            # Calculate angle in degrees
            angle_rad = np.arctan2(y_source, x_source)
            doa[i] = np.rad2deg(angle_rad)
        else:
            doa[i] = 0
    
    # Calculate distances from source to each ear
    # Using the spherical head model from the MATLAB code
    # Left ear
    dist_to_left_ear = np.zeros(total_samples)
    # Right ear
    dist_to_right_ear = np.zeros(total_samples)
    
    for i in range(total_samples):
        # Direct Euclidean distance from source to each ear
        x_source = source_coordinates[i, 0]
        y_source = source_coordinates[i, 1]
        
        # Distance to left ear
        dist_to_left_ear[i] = np.sqrt((x_source - coo_ear[0, 0])**2 + (y_source - coo_ear[0, 1])**2)
        
        # Distance to right ear
        dist_to_right_ear[i] = np.sqrt((x_source - coo_ear[1, 0])**2 + (y_source - coo_ear[1, 1])**2)
    
    # 3. Apply amplitude and delay based on distances
    # Calculate amplitude modulation using inverse square law with attenuation factor
    # This is adapted from the MATLAB code's approach
    ref_level = 100  # Reference sound level
    min_distance = 0.001  # To avoid division by zero
    
    # Use the reference amplitude and attenuation factor
    ref_amplitude = 2e-5 * 10**((ref_level-3)/20)
    
    # Calculate amplitude scaling based on distance
    left_amp = np.zeros(total_samples)
    right_amp = np.zeros(total_samples)
    
    for i in range(total_samples):
        # Amplitude based on distance, with directional bias
        d_left = max(dist_to_left_ear[i], min_distance)
        d_right = max(dist_to_right_ear[i], min_distance)
        d_center = max(distance_to_head_center[i], min_distance)
        
        # Calculate amplitude attenuation
        # Directional component based on angle (adapted from MATLAB)
        a_L = 5/180
        b_L = -5
        loss_dB_L = a_L * doa[i] + b_L
        
        a_R = -5/180
        b_R = 0
        loss_dB_R = a_R * doa[i] + b_R
        
        # Combine direction and distance effects
        left_amp[i] = 2e-5 * 10**((loss_dB_L + DISTANCE_FACTOR * np.log10(1/d_center))/20)
        right_amp[i] = 2e-5 * 10**((loss_dB_R + DISTANCE_FACTOR * np.log10(1/d_center))/20)
    
    # Normalize and scale
    left_amp = left_amp / np.max(left_amp) * ref_amplitude
    right_amp = right_amp / np.max(right_amp) * ref_amplitude
    
    # 4. Calculate time delays based on distance differences
    # Convert distance difference to time delay in samples
    left_delay_samples = np.round(dist_to_left_ear / SPEED_OF_SOUND * SAMPLE_RATE).astype(int)
    right_delay_samples = np.round(dist_to_right_ear / SPEED_OF_SOUND * SAMPLE_RATE).astype(int)
    
    # Find maximum delay to ensure we have enough buffer
    max_delay = max(np.max(left_delay_samples), np.max(right_delay_samples))
    
    # 5. Create output buffer with extra space for delays
    output_signal = np.zeros((total_samples + max_delay, 2))
    
    # 6. Apply the amplitude and delay modulation to the directional signal
    for i in range(total_samples):
        # Apply amplitude scaling and add to output with delay
        output_signal[left_delay_samples[i] + i, 0] += dir_left[i] * left_amp[i]
        output_signal[right_delay_samples[i] + i, 1] += dir_right[i] * right_amp[i]
    
    # Ensure we have enough samples and trim if needed
    if len(output_signal) > total_samples:
        output_signal = output_signal[:total_samples]
    
    # 7. Apply a gentle fade out at the end to avoid any clicks
    fade_samples = int(0.1 * SAMPLE_RATE)
    if fade_samples < total_samples:
        fade_window = np.ones(total_samples)
        fade_window[-fade_samples:] = np.linspace(1.0, 0.0, fade_samples)
        
        output_signal[:, 0] *= fade_window
        output_signal[:, 1] *= fade_window
    
    # 8. Normalize to avoid clipping
    max_val = np.max(np.abs(output_signal))
    if max_val > 0:
        output_signal = output_signal / max_val * 0.9
    
    # Create distance array for plotting
    distances = np.sqrt(x_positions**2 + y_positions**2)
    
    return output_signal, distances

def plot_distance_trajectory(distances, azimuth, output_path):
    """Plot the trajectory of the approaching sound (distance over time)"""
    # Create figure
    plt.figure(figsize=(10, 6))
    
    # Define time axis (in seconds)
    time_axis = np.linspace(0, DURATION, len(distances))
    
    # Plot distance vs time
    plt.plot(time_axis, distances, 'b-', linewidth=3)
    
    # Scatter plot with color gradient representing distance
    scatter_main = plt.scatter(time_axis, distances, c=distances, cmap='viridis', s=80, alpha=0.7)
    
    # Add colorbar to explain the gradient colors
    cbar_main = plt.colorbar(scatter_main)
    cbar_main.set_label('Distance (meters)', fontsize=10)
    
    # Mark start and end points
    plt.scatter(time_axis[0], distances[0], color='green', s=120, marker='o', label="Start (Far)")
    plt.scatter(time_axis[-1], distances[-1], color='red', s=120, marker='o', label="End (Near)")
    
    # Add grid and labels
    plt.grid(True, linestyle='--', alpha=0.5)
    
    # Create azimuth direction string
    if azimuth == 0:
        direction = "Front"
    elif azimuth == 90:
        direction = "Right"
    elif azimuth == -90:
        direction = "Left"
    else:
        direction = f"{azimuth}°"
    
    plt.title(f"Sound Source Approaching from the {direction} ({azimuth}° azimuth)", fontsize=14)
    plt.xlabel("Time (seconds)", fontsize=12)
    plt.ylabel("Distance from Listener (meters)", fontsize=12)
    plt.legend(fontsize=10)
    
    # Add annotations
    plt.annotate("Far", (time_axis[0], distances[0]), 
                xytext=(time_axis[0]-0.2, distances[0]+0.2),
                fontsize=12, fontweight='bold')
    plt.annotate("Near", (time_axis[-1], distances[-1]), 
                xytext=(time_axis[-1]-0.2, distances[-1]+0.2),
                fontsize=12, fontweight='bold')
    
    # Add top-down view of the scene - positioned higher to avoid interfering with axis
    ax_inset = plt.axes([0.15, 0.25, 0.3, 0.3])  # [left, bottom, width, height] - moved higher
    
    # Plot listener (center) and sound positions
    ax_inset.scatter([0], [0], color='black', s=200, marker='o', label="Listener")
    
    # Calculate positions along azimuth at different distances
    # We rotate by 90° to convert from azimuth to Cartesian
    x_positions = distances * np.cos(np.deg2rad(azimuth + 90))
    y_positions = distances * np.sin(np.deg2rad(azimuth + 90))
    
    # Plot trajectory points with color gradient
    scatter_inset = ax_inset.scatter(x_positions, y_positions, c=distances, cmap='viridis', s=50, alpha=0.7)
    ax_inset.plot(x_positions, y_positions, 'k-', alpha=0.5)
    
    # Add colorbar to inset plot
    cbar_inset = plt.colorbar(scatter_inset, ax=ax_inset, shrink=0.7)
    cbar_inset.set_label('Distance (m)', fontsize=8)
    
    # Mark start and end
    ax_inset.scatter(x_positions[0], y_positions[0], color='green', s=80, marker='o')
    ax_inset.scatter(x_positions[-1], y_positions[-1], color='red', s=80, marker='o')
    
    # Add an arrow to show direction of movement
    arrow_idx = len(distances) // 2
    ax_inset.arrow(x_positions[arrow_idx], y_positions[arrow_idx], 
                  (x_positions[arrow_idx+1] - x_positions[arrow_idx])*3, 
                  (y_positions[arrow_idx+1] - y_positions[arrow_idx])*3,
                  head_width=0.1, head_length=0.15, fc='red', ec='red')
    
    # Draw head
    head_circle = plt.Circle((0, 0), HEAD_DIAMETER/2, fill=False, color='blue', alpha=0.7)
    ax_inset.add_artist(head_circle)
    
    # Draw ears
    ax_inset.scatter([-HEAD_DIAMETER/2, HEAD_DIAMETER/2], [0, 0], color='blue', s=50, marker='o')
    
    # Set equal aspect ratio for inset
    ax_inset.set_aspect('equal')
    ax_inset.set_title("Top-Down View", fontsize=10)
    ax_inset.grid(True, linestyle='--', alpha=0.3)
    ax_inset.set_xlabel("X (meters)")
    ax_inset.set_ylabel("Y (meters)")
    
    # Add legend to bottom left of inset plot
    ax_inset.legend(loc='lower left', fontsize=8)
    
    # Set axis limits to show the geometry
    max_range = max(abs(INITIAL_DISTANCE), abs(FINAL_DISTANCE)) * 1.2
    ax_inset.set_xlim(-max_range, max_range)
    ax_inset.set_ylim(-max_range, max_range)
    
    # Save figure
    plt.savefig(output_path, dpi=300, bbox_inches='tight')
    plt.close()
    
    print(f"Saved trajectory plot to: {output_path}")

def main():
    """Generate approaching sound from different directions using Fabian SOFA models with physics-based approach"""
    if not SOFA_AVAILABLE:
        print("ERROR: sofar library is required. Please install it with:")
        print("pip install sofar")
        return
    
    # Generate source signal (mono)
    print("Generating source signal...")
    source_signal = generate_pink_noise(DURATION, SAMPLE_RATE)
    
    # Process each SOFA file
    valid_files = []
    
    for sofa_path in sofa_files:
        if not os.path.exists(sofa_path):
            print(f"File not found: {sofa_path}")
            continue
        
        valid_files.append(sofa_path)
        file_name = os.path.basename(sofa_path).replace('.sofa', '')
        print(f"\nProcessing {file_name}...")
        
        try:
            # Load SOFA file
            hrtf, dimensions = load_sofa_file(sofa_path)
            
            if hrtf is None:
                print(f"Failed to load {file_name}, skipping.")
                continue
            
            # Create approaching movement for each azimuth
            for azimuth in AZIMUTH_ANGLES:
                print(f"\nGenerating approaching sound from azimuth {azimuth}°...")
                
                # Create approaching movement with physics-based model
                output_signal, distances = create_approaching_movement_physics(hrtf, source_signal, azimuth)
                
                # Save audio file
                if azimuth == 0:
                    direction = "front"
                elif azimuth == 90:
                    direction = "right"
                elif azimuth == -90:
                    direction = "left"
                else:
                    direction = f"az{azimuth}"
                
                output_filename = f"approaching_{direction}_az{azimuth}_FABIAN_HRIR_natural.wav"
                output_path = os.path.join(output_dir, output_filename)
                sf.write(output_path, output_signal, SAMPLE_RATE)
                print(f"Saved audio to: {output_path}")
                
                # Create trajectory plot for each azimuth
                plot_path = os.path.join(output_dir, 
                                       f"trajectory_{direction}_az{azimuth}_{file_name}.png")
                plot_distance_trajectory(distances, azimuth, plot_path)
            
        except Exception as e:
            print(f"Error processing {file_name}: {e}")
            import traceback
            traceback.print_exc()
    
    print("\nApproaching movement processing complete!")
    print(f"Total SOFA models processed: {len(valid_files)}")
    print(f"All files saved to: {output_dir}")
    
    # Summarize settings used
    print("\nSettings used:")
    print(f"  Initial distance: {INITIAL_DISTANCE}m")
    print(f"  Final distance: {FINAL_DISTANCE}m")
    print(f"  Motion duration: {DURATION}s")
    print(f"  Azimuths: {AZIMUTH_ANGLES}")
    print(f"  Elevation: {CONSTANT_ELEVATION}°")
    print(f"  Head diameter: {HEAD_DIAMETER}m (natural human head size)")
    print(f"  Distance attenuation factor: {DISTANCE_FACTOR}")
    print(f"  HRTF model used: FABIAN_HRIR_measured_HATO_0 (for natural binaural cues)")

if __name__ == "__main__":
    main()
```

```python
"""
SOFA HRTF Moving Sound Source Simulator

This script creates binaural audio files simulating a sound source approaching the listener,
using the Fabian HRTF models for directional cues and a physically-based distance model
inspired by the EPFL LEMA approach.
"""
import numpy as np
import scipy.signal as signal
import soundfile as sf
import matplotlib.pyplot as plt
import os
import time
import datetime

# First, ensure required libraries are installed
try:
    import sofar
    SOFA_AVAILABLE = True
except ImportError:
    print("Warning: sofar library not found. Please install with: pip install sofar")
    SOFA_AVAILABLE = False

#############################################################
# CONTROL PARAMETERS - ADJUST THESE TO CHANGE THE SIMULATION
#############################################################

# Spatial parameters
INITIAL_DISTANCE = 2.0    # Starting distance in meters
FINAL_DISTANCE = 0.5      # Final distance in meters (increased to avoid extreme effects)
AZIMUTH_ANGLES = [90, 0, -90]  # Angles to generate (right, front, left)
CONSTANT_ELEVATION = 0    # Fixed elevation (ear level)
DISTANCE_FACTOR = 40      # Attenuation factor (higher = stronger distance effect)
HEAD_DIAMETER = 0.18      # Diameter of the head in meters (natural human head size)

# Audio parameters
VOLUME_LEVEL = 1        # Output volume factor (0.0 to 1.0, higher = louder)

# Time parameters
DURATION = 3.0            # Duration of movement in seconds
SAMPLE_RATE = 44100       # Audio sample rate in Hz
SPEED_OF_SOUND = 343.0    # Speed of sound in m/s

# Use only HRIR files for maximum binaural effect
sofa_files = [
    r"C:\Users\cogpsy-vrlab\Downloads\HRIRs_neutral_head_orientation_v4\SOFA\FABIAN_HRIR_measured_HATO_0.sofa"
]

# Create output directory with timestamp
timestamp = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
output_dir = os.path.join(r"C:\Users\cogpsy-vrlab\Documents\PPS_module\BreathingPilot\LoomingStimuli", f"looming_stimuli_{timestamp}")
os.makedirs(output_dir, exist_ok=True)

def generate_pink_noise(duration, fs):
    """Generate pink noise using filter-based method (similar to MATLAB code)"""
    # Create filter coefficients for pink noise
    # These are similar to the coefficients used in the MATLAB code
    b = np.array([0.049922035, -0.095993537, 0.050612699, -0.004408786])
    a = np.array([1.0, -2.494956002, 2.017265875, -0.522189400])
    
    # Generate white noise
    np.random.seed(42)  # For reproducibility
    white_noise = np.random.randn(int(duration * fs))
    
    # Apply filter to get pink noise
    pink_noise = signal.lfilter(b, a, white_noise)
    
    # Normalize
    pink_noise = pink_noise / np.max(np.abs(pink_noise)) * 0.8
    
    return pink_noise

def load_sofa_file(sofa_path):
    """Load HRTF data from SOFA file"""
    try:
        print(f"Loading SOFA file: {os.path.basename(sofa_path)}")
        
        # Load the SOFA file
        hrtf = sofar.read_sofa(sofa_path)
        
        # Display basic SOFA info
        print(f"SOFA convention: {hrtf.GLOBAL_SOFAConventions} v{hrtf.GLOBAL_SOFAConventionsVersion}")
        
        # Get dimensions
        dimensions = {}
        for dim_name in ['M', 'R', 'N']:
            try:
                dimensions[dim_name] = hrtf.get_dimension(dim_name)
                print(f"{dim_name} = {dimensions[dim_name]}")
            except:
                dimensions[dim_name] = None
                print(f"{dim_name} = Unknown")
        
        return hrtf, dimensions
        
    except Exception as e:
        print(f"Error loading SOFA file: {e}")
        import traceback
        traceback.print_exc()
        return None, None

def find_nearest_hrtf(hrtf, target_azimuth, target_elevation=0, target_distance=1.0):
    """Find the measurement closest to the target azimuth, elevation, and distance"""
    try:
        # Check if this is a CTF file (Common Transfer Function)
        # CTF files typically have only one measurement/position
        is_ctf = False
        if "CTF" in hrtf.GLOBAL_SOFAConventions or "CTF" in str(hrtf):
            is_ctf = True
            
        if is_ctf or (hasattr(hrtf, 'Data_IR') and hrtf.Data_IR.shape[0] == 1 and hrtf.Data_IR.shape[1] == 1):
            print(f"Detected CTF file - using the single filter for azimuth {target_azimuth}°")
            # For CTF, there's only one measurement so return index 0
            # Try to get the position if available, otherwise use default
            if hasattr(hrtf, 'SourcePosition') and len(hrtf.SourcePosition) > 0:
                nearest_pos = hrtf.SourcePosition[0] if len(hrtf.SourcePosition.shape) == 1 else hrtf.SourcePosition[0,:]
            else:
                nearest_pos = np.array([0.0, 0.0, 0.0])  # Default position
            return 0, nearest_pos
    
        # Get source positions for regular HRTF files
        source_positions = np.array(hrtf.SourcePosition)
        
        # Handle different position array shapes
        if source_positions.shape[0] == 3 and len(source_positions.shape) == 2:
            source_positions = source_positions.T
        
        # Check if positions are in degrees or radians
        position_type = hrtf.SourcePosition_Type
        max_azimuth = np.max(np.abs(source_positions[:, 0]))
        positions_in_degrees = max_azimuth > 6.28
        
        if positions_in_degrees:
            # Both source positions and target are in degrees
            az_compare = target_azimuth
            el_compare = target_elevation
        else:
            # Source positions in radians, convert target to radians
            az_compare = np.deg2rad(target_azimuth)
            el_compare = np.deg2rad(target_elevation)
        
        # Handle negative azimuths if data is in 0-360 format
        if positions_in_degrees and target_azimuth < 0:
            # Check if we have negative azimuths in the data
            has_negative_az = np.any(source_positions[:, 0] < 0)
            if not has_negative_az:
                # Convert target from -180 to 180 format to 0 to 360 format
                az_compare = (target_azimuth + 360) % 360
        
        # Create distance metric focusing on azimuth and elevation
        if position_type == 'spherical':
            dists = np.sqrt(
                2 * (source_positions[:, 0] - az_compare)**2 + 
                (source_positions[:, 1] - el_compare)**2
            )
        else:
            # For Cartesian or unknown position type
            dists = np.abs(source_positions[:, 0] - az_compare)
                
        # Find nearest position
        nearest_idx = np.argmin(dists)
        nearest_pos = source_positions[nearest_idx]
        
        return nearest_idx, nearest_pos
        
    except Exception as e:
        print(f"Error finding nearest HRTF: {e}")
        import traceback
        traceback.print_exc()
        return 0, None

def extract_hrtf_data(hrtf, idx):
    """Extract HRTF impulse responses for a given index"""
    hrtf_data = hrtf.Data_IR
    
    # Check if this is a CTF file (Common Transfer Function)
    # CTF files often have shape (1, 1, N) or similar with only one filter
    is_ctf = False
    if "CTF" in hrtf.GLOBAL_SOFAConventions or "CTF" in str(hrtf):
        is_ctf = True
    
    # Handle CTF files (apply same filter to both ears)
    if is_ctf or hrtf_data.shape[0] == 1 and hrtf_data.shape[1] == 1:
        print("Detected CTF file - using same filter for both ears")
        try:
            # Get the single filter
            common_filter = hrtf_data[0, 0, :]
            return common_filter, common_filter
        except Exception as e:
            print(f"Error extracting CTF data: {e}")
            # Return simple impulse responses as fallback
            dummy_ir = np.zeros(256)
            dummy_ir[0] = 1.0
            return dummy_ir, dummy_ir
    
    # Regular HRTF files with separate left/right filters
    # Check dimensions and extract
    if hrtf_data.shape[0] == 2 and hrtf_data.shape[1] > 2:
        # Shape is (R, M, N)
        hrtf_left = hrtf_data[0, idx, :]
        hrtf_right = hrtf_data[1, idx, :]
    else:
        # Shape is (M, R, N)
        try:
            hrtf_left = hrtf_data[idx, 0, :]
            hrtf_right = hrtf_data[idx, 1, :]
        except:
            # Try other dimension orderings if the above fails
            try:
                if len(hrtf_data.shape) == 4:  # If 4D array
                    hrtf_left = hrtf_data[idx, 0, 0, :]
                    hrtf_right = hrtf_data[idx, 1, 0, :]
                else:
                    raise ValueError(f"Unexpected HRTF data shape: {hrtf_data.shape}")
            except Exception as e:
                print(f"Error extracting HRTF data: {e}")
                # Return empty IRs as fallback
                dummy_ir = np.zeros(256)
                dummy_ir[0] = 1.0
                return dummy_ir, dummy_ir
    
    return hrtf_left, hrtf_right

def create_approaching_movement_physics(hrtf, source_signal, azimuth, verbose=True):
    """
    Create audio where the sound approaches from a fixed azimuth using the EPFL LEMA
    physically-based approach for distance effects
    """
    # Calculate total samples
    total_samples = int(DURATION * SAMPLE_RATE)
    
    # Use HRTF for directional cues - find closest matching HRTF for our azimuth
    nearest_idx, nearest_pos = find_nearest_hrtf(hrtf, azimuth, CONSTANT_ELEVATION)
    hrtf_left, hrtf_right = extract_hrtf_data(hrtf, nearest_idx)
    
    if verbose:
        if nearest_pos is not None:
            try:
                # For spherical coordinates
                if hasattr(hrtf, 'SourcePosition_Type') and hrtf.SourcePosition_Type == 'spherical':
                    # Check if in degrees or radians
                    if isinstance(nearest_pos, np.ndarray) and len(nearest_pos) > 0:
                        if np.max(np.abs(nearest_pos[0])) > 6.28:
                            pos_str = f"({nearest_pos[0]:.1f}°, {nearest_pos[1]:.1f}°)"
                        else:
                            pos_str = f"({np.rad2deg(nearest_pos[0]):.1f}°, {np.rad2deg(nearest_pos[1]):.1f}°)"
                    else:
                        pos_str = str(nearest_pos)
                else:
                    # For Cartesian or other formats
                    if isinstance(nearest_pos, np.ndarray) and len(nearest_pos) >= 3:
                        pos_str = f"[{nearest_pos[0]:.2f}, {nearest_pos[1]:.2f}, {nearest_pos[2]:.2f}]"
                    else:
                        pos_str = str(nearest_pos)
            except:
                pos_str = str(nearest_pos)
        else:
            pos_str = "unknown"
            
        print(f"Using HRTF measurement at position {pos_str}")
    
    # Apply the HRTF to the entire source signal for directional cues
    print("Applying HRTF filter for directional cues...")
    dir_left = signal.fftconvolve(source_signal, hrtf_left, mode='same')
    dir_right = signal.fftconvolve(source_signal, hrtf_right, mode='same')
    
    # Now apply the physics-based approach from the MATLAB code for distance effects
    
    # 1. Create source coordinates as it approaches, similar to the MATLAB code
    # Set up the geometry (x, y) coordinates
    # Coordinates of the ears (left and right)
    coo_ear = np.array([[-HEAD_DIAMETER/2, 0], [HEAD_DIAMETER/2, 0]])
    
    # Calculate the y-coordinate based on azimuth
    # For straight approaches, y is constant, x changes
    if azimuth == 0:  # Coming from front
        x_start = -INITIAL_DISTANCE
        x_stop = -FINAL_DISTANCE
        y_value = 0
    elif azimuth == 90:  # Coming from right
        x_start = 0
        x_stop = 0
        y_value = INITIAL_DISTANCE
        y_stop = FINAL_DISTANCE
    elif azimuth == -90:  # Coming from left
        x_start = 0 
        x_stop = 0
        y_value = -INITIAL_DISTANCE
        y_stop = -FINAL_DISTANCE
    else:
        # For other angles, calculate the x and y components
        angle_rad = np.deg2rad(azimuth + 90)  # Convert to radians with 0 = front
        x_start = INITIAL_DISTANCE * np.cos(angle_rad)
        y_start = INITIAL_DISTANCE * np.sin(angle_rad)
        x_stop = FINAL_DISTANCE * np.cos(angle_rad)
        y_stop = FINAL_DISTANCE * np.sin(angle_rad)
    
    # Create source coordinates for each sample
    time_points = np.linspace(0, DURATION, total_samples)
    if azimuth == 90 or azimuth == -90:
        # For right or left approaches, x is constant, y changes
        x_positions = np.full(total_samples, x_start)
        if azimuth == 90:  # Right side
            y_positions = np.linspace(y_value, y_stop, total_samples)
        else:  # Left side
            y_positions = np.linspace(y_value, y_stop, total_samples)
    else:
        # For other angles
        x_positions = np.linspace(x_start, x_stop, total_samples)
        if azimuth == 0:  # Front
            y_positions = np.full(total_samples, y_value)
        else:
            y_positions = np.linspace(y_start, y_stop, total_samples)
    
    # Combine into source coordinates
    source_coordinates = np.column_stack((x_positions, y_positions))
    
    # 2. Calculate distances from the source to each ear at each time point
    # Distance from source to head center
    distance_to_head_center = np.sqrt(source_coordinates[:, 0]**2 + source_coordinates[:, 1]**2)
    
    # Calculate the direction of arrival angle for interaural time difference
    doa = np.zeros(total_samples)
    for i in range(total_samples):
        x_source = source_coordinates[i, 0]
        y_source = source_coordinates[i, 1]
        source_distance = np.sqrt(x_source**2 + y_source**2)
        
        if source_distance > 0:
            # Calculate angle in degrees
            angle_rad = np.arctan2(y_source, x_source)
            doa[i] = np.rad2deg(angle_rad)
        else:
            doa[i] = 0
    
    # Calculate distances from source to each ear
    # Using the spherical head model from the MATLAB code
    # Left ear
    dist_to_left_ear = np.zeros(total_samples)
    # Right ear
    dist_to_right_ear = np.zeros(total_samples)
    
    for i in range(total_samples):
        # Direct Euclidean distance from source to each ear
        x_source = source_coordinates[i, 0]
        y_source = source_coordinates[i, 1]
        
        # Distance to left ear
        dist_to_left_ear[i] = np.sqrt((x_source - coo_ear[0, 0])**2 + (y_source - coo_ear[0, 1])**2)
        
        # Distance to right ear
        dist_to_right_ear[i] = np.sqrt((x_source - coo_ear[1, 0])**2 + (y_source - coo_ear[1, 1])**2)
    
    # 3. Apply amplitude and delay based on distances
    # Calculate amplitude modulation using inverse square law with attenuation factor
    # This is adapted from the MATLAB code's approach
    ref_level = 100  # Reference sound level
    min_distance = 0.001  # To avoid division by zero
    
    # Use the reference amplitude and attenuation factor
    ref_amplitude = 2e-5 * 10**((ref_level-3)/20)
    
    # Calculate amplitude scaling based on distance
    left_amp = np.zeros(total_samples)
    right_amp = np.zeros(total_samples)
    
    for i in range(total_samples):
        # Amplitude based on distance, with directional bias
        d_left = max(dist_to_left_ear[i], min_distance)
        d_right = max(dist_to_right_ear[i], min_distance)
        d_center = max(distance_to_head_center[i], min_distance)
        
        # Calculate amplitude attenuation
        # Directional component based on angle (adapted from MATLAB)
        a_L = 5/180
        b_L = -5
        loss_dB_L = a_L * doa[i] + b_L
        
        a_R = -5/180
        b_R = 0
        loss_dB_R = a_R * doa[i] + b_R
        
        # Combine direction and distance effects
        left_amp[i] = 2e-5 * 10**((loss_dB_L + DISTANCE_FACTOR * np.log10(1/d_center))/20)
        right_amp[i] = 2e-5 * 10**((loss_dB_R + DISTANCE_FACTOR * np.log10(1/d_center))/20)
    
    # Normalize and scale
    left_amp = left_amp / np.max(left_amp) * ref_amplitude
    right_amp = right_amp / np.max(right_amp) * ref_amplitude
    
    # 4. Calculate time delays based on distance differences
    # Convert distance difference to time delay in samples
    left_delay_samples = np.round(dist_to_left_ear / SPEED_OF_SOUND * SAMPLE_RATE).astype(int)
    right_delay_samples = np.round(dist_to_right_ear / SPEED_OF_SOUND * SAMPLE_RATE).astype(int)
    
    # Find maximum delay to ensure we have enough buffer
    max_delay = max(np.max(left_delay_samples), np.max(right_delay_samples))
    
    # 5. Create output buffer with extra space for delays
    output_signal = np.zeros((total_samples + max_delay, 2))
    
    # 6. Apply the amplitude and delay modulation to the directional signal
    for i in range(total_samples):
        # Apply amplitude scaling and add to output with delay
        output_signal[left_delay_samples[i] + i, 0] += dir_left[i] * left_amp[i]
        output_signal[right_delay_samples[i] + i, 1] += dir_right[i] * right_amp[i]
    
    # Ensure we have enough samples and trim if needed
    if len(output_signal) > total_samples:
        output_signal = output_signal[:total_samples]
    
    # 7. Apply a gentle fade out at the end to avoid any clicks
    fade_samples = int(0.1 * SAMPLE_RATE)
    if fade_samples < total_samples:
        fade_window = np.ones(total_samples)
        fade_window[-fade_samples:] = np.linspace(1.0, 0.0, fade_samples)
        
        output_signal[:, 0] *= fade_window
        output_signal[:, 1] *= fade_window
    
    # 8. Normalize to avoid clipping and apply volume control
    max_val = np.max(np.abs(output_signal))
    if max_val > 0:
        # First normalize to avoid clipping
        output_signal = output_signal / max_val * 0.9
        
        # Then apply the volume factor
        output_signal = output_signal * VOLUME_LEVEL
    
    # Create distance array for plotting
    distances = np.sqrt(x_positions**2 + y_positions**2)
    
    return output_signal, distances

def plot_distance_trajectory(distances, azimuth, output_path):
    """Plot the trajectory of the approaching sound (distance over time)"""
    # Create figure
    plt.figure(figsize=(10, 6))
    
    # Define time axis (in seconds)
    time_axis = np.linspace(0, DURATION, len(distances))
    
    # Plot distance vs time
    plt.plot(time_axis, distances, 'b-', linewidth=3)
    
    # Scatter plot with color gradient representing distance
    scatter_main = plt.scatter(time_axis, distances, c=distances, cmap='viridis', s=80, alpha=0.7)
    
    # Add colorbar to explain the gradient colors
    cbar_main = plt.colorbar(scatter_main)
    cbar_main.set_label('Distance (meters)', fontsize=10)
    
    # Mark start and end points
    plt.scatter(time_axis[0], distances[0], color='green', s=120, marker='o', label="Start (Far)")
    plt.scatter(time_axis[-1], distances[-1], color='red', s=120, marker='o', label="End (Near)")
    
    # Add grid and labels
    plt.grid(True, linestyle='--', alpha=0.5)
    
    # Create azimuth direction string
    if azimuth == 0:
        direction = "Front"
    elif azimuth == 90:
        direction = "Right"
    elif azimuth == -90:
        direction = "Left"
    else:
        direction = f"{azimuth}°"
    
    plt.title(f"Sound Source Approaching from the {direction} ({azimuth}° azimuth)", fontsize=14)
    plt.xlabel("Time (seconds)", fontsize=12)
    plt.ylabel("Distance from Listener (meters)", fontsize=12)
    plt.legend(fontsize=10)
    
    # Add annotations
    plt.annotate("Far", (time_axis[0], distances[0]), 
                xytext=(time_axis[0]-0.2, distances[0]+0.2),
                fontsize=12, fontweight='bold')
    plt.annotate("Near", (time_axis[-1], distances[-1]), 
                xytext=(time_axis[-1]-0.2, distances[-1]+0.2),
                fontsize=12, fontweight='bold')
    
    # Add top-down view of the scene - positioned higher to avoid interfering with axis
    ax_inset = plt.axes([0.15, 0.25, 0.3, 0.3])  # [left, bottom, width, height] - moved higher
    
    # Plot listener (center) and sound positions
    ax_inset.scatter([0], [0], color='black', s=200, marker='o', label="Listener")
    
    # Calculate positions along azimuth at different distances
    # We rotate by 90° to convert from azimuth to Cartesian
    x_positions = distances * np.cos(np.deg2rad(azimuth + 90))
    y_positions = distances * np.sin(np.deg2rad(azimuth + 90))
    
    # Plot trajectory points with color gradient
    scatter_inset = ax_inset.scatter(x_positions, y_positions, c=distances, cmap='viridis', s=50, alpha=0.7)
    ax_inset.plot(x_positions, y_positions, 'k-', alpha=0.5)
    
    # Add colorbar to inset plot
    cbar_inset = plt.colorbar(scatter_inset, ax=ax_inset, shrink=0.7)
    cbar_inset.set_label('Distance (m)', fontsize=8)
    
    # Mark start and end
    ax_inset.scatter(x_positions[0], y_positions[0], color='green', s=80, marker='o')
    ax_inset.scatter(x_positions[-1], y_positions[-1], color='red', s=80, marker='o')
    
    # Add an arrow to show direction of movement
    arrow_idx = len(distances) // 2
    ax_inset.arrow(x_positions[arrow_idx], y_positions[arrow_idx], 
                  (x_positions[arrow_idx+1] - x_positions[arrow_idx])*3, 
                  (y_positions[arrow_idx+1] - y_positions[arrow_idx])*3,
                  head_width=0.1, head_length=0.15, fc='red', ec='red')
    
    # Draw head
    head_circle = plt.Circle((0, 0), HEAD_DIAMETER/2, fill=False, color='blue', alpha=0.7)
    ax_inset.add_artist(head_circle)
    
    # Draw ears
    ax_inset.scatter([-HEAD_DIAMETER/2, HEAD_DIAMETER/2], [0, 0], color='blue', s=50, marker='o')
    
    # Set equal aspect ratio for inset
    ax_inset.set_aspect('equal')
    ax_inset.set_title("Top-Down View", fontsize=10)
    ax_inset.grid(True, linestyle='--', alpha=0.3)
    ax_inset.set_xlabel("X (meters)")
    ax_inset.set_ylabel("Y (meters)")
    
    # Add legend to bottom left of inset plot
    ax_inset.legend(loc='lower left', fontsize=8)
    
    # Set axis limits to show the geometry
    max_range = max(abs(INITIAL_DISTANCE), abs(FINAL_DISTANCE)) * 1.2
    ax_inset.set_xlim(-max_range, max_range)
    ax_inset.set_ylim(-max_range, max_range)
    
    # Save figure
    plt.savefig(output_path, dpi=300, bbox_inches='tight')
    plt.close()
    
    print(f"Saved trajectory plot to: {output_path}")

def main():
    """Generate approaching sound from different directions using Fabian SOFA models with physics-based approach"""
    if not SOFA_AVAILABLE:
        print("ERROR: sofar library is required. Please install it with:")
        print("pip install sofar")
        return
    
    # Generate source signal (mono)
    print("Generating source signal...")
    source_signal = generate_pink_noise(DURATION, SAMPLE_RATE)
    
    # Process each SOFA file
    valid_files = []
    
    for sofa_path in sofa_files:
        if not os.path.exists(sofa_path):
            print(f"File not found: {sofa_path}")
            continue
        
        valid_files.append(sofa_path)
        file_name = os.path.basename(sofa_path).replace('.sofa', '')
        print(f"\nProcessing {file_name}...")
        
        try:
            # Load SOFA file
            hrtf, dimensions = load_sofa_file(sofa_path)
            
            if hrtf is None:
                print(f"Failed to load {file_name}, skipping.")
                continue
            
            # Create approaching movement for each azimuth
            for azimuth in AZIMUTH_ANGLES:
                print(f"\nGenerating approaching sound from azimuth {azimuth}°...")
                
                # Create approaching movement with physics-based model
                output_signal, distances = create_approaching_movement_physics(hrtf, source_signal, azimuth)
                
                # Save audio file
                if azimuth == 0:
                    direction = "front"
                elif azimuth == 90:
                    direction = "right"
                elif azimuth == -90:
                    direction = "left"
                else:
                    direction = f"az{azimuth}"
                
                output_filename = f"approaching_{direction}_az{azimuth}_FABIAN_HRIR_natural.wav"
                output_path = os.path.join(output_dir, output_filename)
                sf.write(output_path, output_signal, SAMPLE_RATE)
                print(f"Saved audio to: {output_path}")
                
                # Create trajectory plot for each azimuth
                plot_path = os.path.join(output_dir, 
                                       f"trajectory_{direction}_az{azimuth}_{file_name}.png")
                plot_distance_trajectory(distances, azimuth, plot_path)
            
        except Exception as e:
            print(f"Error processing {file_name}: {e}")
            import traceback
            traceback.print_exc()
    
    print("\nApproaching movement processing complete!")
    print(f"Total SOFA models processed: {len(valid_files)}")
    print(f"All files saved to: {output_dir}")
    
    # Summarize settings used
    print("\nSettings used:")
    print(f"  Initial distance: {INITIAL_DISTANCE}m")
    print(f"  Final distance: {FINAL_DISTANCE}m")
    print(f"  Motion duration: {DURATION}s")
    print(f"  Azimuths: {AZIMUTH_ANGLES}")
    print(f"  Elevation: {CONSTANT_ELEVATION}°")
    print(f"  Head diameter: {HEAD_DIAMETER}m (natural human head size)")
    print(f"  Distance attenuation factor: {DISTANCE_FACTOR}")
    print(f"  Volume level: {VOLUME_LEVEL:.1f} (0.0-1.0)")
    print(f"  HRTF model used: FABIAN_HRIR_measured_HATO_0 (for natural binaural cues)")

if __name__ == "__main__":
    main()
```
