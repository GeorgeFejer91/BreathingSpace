```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-
"""
PPS Design Generator

This script generates participant‐specific experimental design CSVs for a 
Peripersonal Space (PPS) experiment. Each design CSV is built upon a 
"scaffold" CSV that contains the retention slot (breathhold) timestamps 
from a box‐breathing audio file. If needed, additional retention slots are 
appended (using fixed 8‑second increments) so that every trial can be assigned 
to a specific retention slot. Then, a counterbalanced set of trial conditions 
(e.g., trial type, looming stimulus type, SOA, jitter) is allocated to each 
scaffold row according to the following rules:
  
  - Inhalation trials (looming cue + tactile) are only assigned to scaffold rows 
    whose breathphase is "inhale."
  - Exhalation trials (looming cue + tactile) are only assigned to scaffold rows 
    whose breathphase is "exhale."
  - Baseline (tactile only; looming = NA) and catch (looming only; no tactile) trials 
    are allocated from the remaining pool in a random, counterbalanced manner 
    between the inhale and exhale rows.
    
The final output retains the original column order from the scaffold, with new 
trial‐specific variables appended as new columns.

Author: AI Assistant
"""

import os
import numpy as np
import pandas as pd
import random
import traceback

# ----------------------------------------------------------------------
# Helper function: Convert milliseconds to "MM:SS.s" format
# ----------------------------------------------------------------------
def convert_ms_to_timestamp(ms):
    m = int(ms // 60000)
    s = (ms % 60000) / 1000.0
    return f"{m:02}:{s:.1f}"

# ----------------------------------------------------------------------
# CONFIGURATION PARAMETERS
# ----------------------------------------------------------------------
CONFIG = {
    'repetitions': 12,           # Number of trials per SOA condition
    'num_participants': 100,     # Default number of participants
    'practice_trials': 8,        # Number of practice trials at the beginning
    'included_conditions': {
        'inhalation': True,
        'exhalation': True,
        'baseline': True,
    },
    'catch_trial_percentage': 10,
    'soa_conditions_ms': [190, 400, 700, 1000, 1500],
    'jitter_options_ms': [100, 200, 300, 400, 500],
    'debug_mode': True,
    'paths': {
        # Directory where participant-specific CSV files will be saved
        'experiment_log_dir': r"C:\Users\cogpsy-vrlab\Documents\GitHub\BreathingSpace\Level1_AudioGeneration\ExperimentLog",
    },
    # Path to the original scaffold CSV containing retention (breathhold) timestamps.
    'skeleton_csv_path': r"C:\Users\cogpsy-vrlab\Documents\GitHub\BreathingSpace\Level0_StimulusFiles\BoxBreathing_timestamps.csv",
    'box_breathing': {
        'cycle_duration_sec': 8,      # Duration of one breath cycle in seconds
        # Intro/outro are not used here because the scaffold drives timing.
        'intro_duration_sec': 60,
        'outro_duration_sec': 30,
        'sample_rate': 48000,         # Audio sample rate (samples per second)
    }
}

# Noise types for looming stimuli (can be any labels)
LOOMING_STIMULI = ['blue', 'brown', 'pink', 'white']

# ----------------------------------------------------------------------
# CLASS: PPSDesignGenerator
# ----------------------------------------------------------------------
class PPSDesignGenerator:
    """Generates participant-specific CSV design files for PPS experiments."""
    
    def __init__(self, config=None):
        self.config = config or CONFIG
        
        # Prepare output directory.
        self.experiment_log_dir = self.config['paths']['experiment_log_dir']
        os.makedirs(self.experiment_log_dir, exist_ok=True)
        
        # Calculate trial counts.
        # For each condition (inhalation, exhalation, baseline): count = (# noise types) × (# SOA) × (repetitions).
        # Then, add 10% catch trials.
        self.trial_counts = self._calculate_trial_counts()
        self.conditions = [cond for cond, inc in self.config['included_conditions'].items() if inc]
        
        # Set seed for reproducibility.
        self.base_seed = 42
        
        if self.config['debug_mode']:
            total_trials = sum(self.trial_counts.values())
            print("Initialized PPSDesignGenerator with:")
            print(f"  - Repetitions: {self.config['repetitions']}")
            print(f"  - Practice trials: {self.config['practice_trials']}")
            print(f"  - Conditions: {self.conditions}")
            print(f"  - SOA values: {self.config['soa_conditions_ms']}")
            print(f"  - Total trials per participant: {total_trials} + {self.config['practice_trials']} practice = {total_trials + self.config['practice_trials']}")
            print(f"  - Output directory: {self.experiment_log_dir}")
    
    def _calculate_trial_counts(self):
        """
        Calculate the number of trials for each condition, using the repetitions parameter
        to determine how many trials per SOA per condition.
        """
        soa_conditions = self.config['soa_conditions_ms']
        repetitions = self.config['repetitions']  # This now directly controls trials per SOA
        catch_percentage = self.config['catch_trial_percentage']
        
        trial_counts = {}
        for condition, include in self.config['included_conditions'].items():
            if include:
                # Each condition gets repetitions × number of SOAs trials
                trial_counts[condition] = len(soa_conditions) * repetitions
        
        regular_trials = sum(trial_counts.values())
        catch_trials = int(np.ceil(regular_trials * catch_percentage / 100))
        trial_counts['catch'] = catch_trials
        
        if self.config['debug_mode']:
            print("Trial count calculation:")
            for cond, count in trial_counts.items():
                print(f"  {cond}: {count} trials")
            print(f"  Total: {sum(trial_counts.values())} trials")
        
        return trial_counts

    # ------------------------------------------------------------------
    # Load the scaffold CSV and extend it if needed.
    # Retain the original column order.
    # ------------------------------------------------------------------
    def load_retention_slots(self, required_slots):
        """
        Load the scaffold CSV containing retention slots and extend it if needed.
        Handles different column naming variations by standardizing them.
        """
        skeleton_path = self.config['skeleton_csv_path']
        try:
            # Try to read the CSV, handling different separators
            skeleton_df = pd.read_csv(skeleton_path, sep=None, engine='python')
            if len(skeleton_df.columns) == 1:
                skeleton_df = pd.read_csv(skeleton_path, delimiter="\t")
            
            # Normalize column names (strip whitespace and convert to lowercase)
            skeleton_df.columns = [col.strip().lower() for col in skeleton_df.columns]
            self.scaffold_order = skeleton_df.columns.tolist()
            
            if self.config['debug_mode']:
                print("Scaffold CSV columns:", self.scaffold_order)
                
            # Determine timestamp column name - could be 'timestamp' or 'retentionslot_timestamp'
            timestamp_col = None
            if 'timestamp' in skeleton_df.columns:
                timestamp_col = 'timestamp'
            elif 'retentionslot_timestamp' in skeleton_df.columns:
                timestamp_col = 'retentionslot_timestamp'
            
            # If neither standard name exists, try to find a column that might contain timestamp info
            if timestamp_col is None:
                for col in skeleton_df.columns:
                    if 'time' in col.lower() and not col.lower() == 'milliseconds':
                        timestamp_col = col
                        break
            
            if timestamp_col is None:
                raise ValueError("Could not find timestamp column in scaffold CSV.")
                
            # Standardize the timestamp column name to 'retentionslot_timestamp'
            if timestamp_col != 'retentionslot_timestamp':
                skeleton_df = skeleton_df.rename(columns={timestamp_col: 'retentionslot_timestamp'})
                if timestamp_col in self.scaffold_order:
                    self.scaffold_order[self.scaffold_order.index(timestamp_col)] = 'retentionslot_timestamp'
                
        except Exception as e:
            raise ValueError(f"Error reading scaffold CSV from {skeleton_path}: {e}")
        
        current_slots = len(skeleton_df)
        required_slots = int(required_slots)
        sample_rate = self.config['box_breathing']['sample_rate']
        cycle_sec = self.config['box_breathing']['cycle_duration_sec']
        
        if self.config['debug_mode']:
            print(f"Loaded scaffold CSV with {current_slots} retention slots.")
            # Display the last few existing timestamps
            try:
                print(f"Last timestamp in skeleton CSV: {skeleton_df.iloc[-1]['retentionslot_timestamp']} ({skeleton_df.iloc[-1]['milliseconds']} ms)")
            except Exception as e:
                print(f"Could not display last timestamp: {e}")
        
        if current_slots < required_slots:
            last_row = skeleton_df.iloc[-1]
            last_ms = float(last_row['milliseconds'])
            last_sample_index = int(last_row['sample_index'])
            new_rows = []
            
            # Get the last breathphase to ensure proper alternation
            last_breathphase = last_row['breathphase'].lower()
            
            for i in range(required_slots - current_slots):
                # Increment by 8 seconds (8000 ms) for each new row
                new_ms = last_ms + (cycle_sec * 1000) * (i + 1)
                new_sample_index = last_sample_index + int(cycle_sec * sample_rate * (i + 1))
                new_timestamp = convert_ms_to_timestamp(new_ms)
                
                # Alternate breathphase: if last was 'exhale', next is 'inhale' and vice versa
                if i == 0:  # First new row
                    new_breathphase = 'inhale' if last_breathphase == 'exhale' else 'exhale'
                else:  # Subsequent rows
                    prev_phase = new_rows[i-1]['breathphase'].lower()
                    new_breathphase = 'inhale' if prev_phase == 'exhale' else 'exhale'
                
                new_description = f"Additional hold {i+1}"
                
                # Create the new row with the correct timestamp column name
                new_row = {
                    'retentionslot_timestamp': new_timestamp,
                    'milliseconds': new_ms,
                    'sample_index': new_sample_index,
                    'description': new_description,
                    'breathphase': new_breathphase
                }
                new_rows.append(new_row)
                
                if self.config['debug_mode'] and i < 5:  # Show first few new rows
                    print(f"Adding new row: {new_timestamp} ({new_ms} ms) - {new_breathphase}")
            
            if new_rows:
                new_rows_df = pd.DataFrame(new_rows)
                # Ensure new_rows_df has all columns from skeleton_df
                for col in skeleton_df.columns:
                    if col not in new_rows_df.columns:
                        new_rows_df[col] = None
                
                # Concatenate with only the columns from the original DataFrame
                skeleton_df = pd.concat([skeleton_df, new_rows_df[skeleton_df.columns]], ignore_index=True)
            
            if self.config['debug_mode']:
                print(f"Extended scaffold CSV to {len(skeleton_df)} retention slots.")
                print(f"New timestamp range: {skeleton_df['retentionslot_timestamp'].iloc[0]} to {skeleton_df['retentionslot_timestamp'].iloc[-1]}")
        
        if len(skeleton_df) > required_slots:
            skeleton_df = skeleton_df.iloc[:required_slots]
        
        return skeleton_df

    # ------------------------------------------------------------------
    # Generate counterbalanced stimulus design (trial conditions)
    # ------------------------------------------------------------------
    def generate_counterbalanced_design(self, participant_id):
        """
        Generate a counterbalanced design for a participant with trials per SOA
        determined by the repetitions parameter in the config.
        
        Catch trials don't have SOA values assigned since they don't have tactile stimuli.
        """
        random.seed(self.base_seed + participant_id)
        np.random.seed(self.base_seed + participant_id)
        
        soa_conditions = self.config['soa_conditions_ms']
        jitter_options = self.config['jitter_options_ms']
        stimulus_types = LOOMING_STIMULI
        repetitions = self.config['repetitions']  # Trials per SOA per condition
        
        all_trials = []
        
        # Generate main conditions: inhalation, exhalation, and baseline
        for condition, count in self.trial_counts.items():
            if condition == 'catch' or count == 0:
                continue
            
            # For each SOA, generate exactly 'repetitions' trials
            for soa in soa_conditions:
                # Calculate how many trials per stimulus type (as evenly as possible)
                trials_per_soa = repetitions
                trials_per_stim = trials_per_soa // len(stimulus_types)  # Integer division
                remainder = trials_per_soa % len(stimulus_types)
                
                # Distribute trials across stimulus types
                stim_trials = []
                for i, stim_type in enumerate(stimulus_types):
                    # Add extra trial for remainder distribution
                    extra = 1 if i < remainder else 0
                    for _ in range(trials_per_stim + extra):
                        stim_trials.append(stim_type)
                
                # Shuffle the stimulus types
                random.shuffle(stim_trials)
                
                # Create trials for this SOA
                for stim_type in stim_trials:
                    all_trials.append({
                        'participant_id': participant_id,
                        'trial_type': condition,
                        'stimulus_type': stim_type,
                        'soa_value_ms': soa,
                        'jitter_ms': random.choice(jitter_options),
                        'is_tactile': True  # For these main trials, tactile is present.
                    })
        
        # Add catch trials (looming only, no tactile) - no SOA values needed
        catch_count = self.trial_counts.get('catch', 0)
        if catch_count > 0:
            # Distribute catch trials evenly across stimulus types only
            catch_trials = []
            stim_counts = {stim: 0 for stim in stimulus_types}
            
            for _ in range(catch_count):
                # Choose the stimulus type with the lowest count
                min_stim = min(stim_counts, key=stim_counts.get)
                
                catch_trials.append({
                    'participant_id': participant_id,
                    'trial_type': 'catch',
                    'stimulus_type': min_stim,
                    'soa_value_ms': None,  # No SOA for catch trials
                    'jitter_ms': random.choice(jitter_options),
                    'is_tactile': False
                })
                
                stim_counts[min_stim] += 1
            
            # Shuffle catch trials and add to all trials
            random.shuffle(catch_trials)
            all_trials.extend(catch_trials)
        
        # Shuffle all trials and assign trial numbers
        all_trials_df = pd.DataFrame(all_trials)
        all_trials_df = all_trials_df.sample(frac=1, random_state=participant_id*100).reset_index(drop=True)
        all_trials_df['trial_number'] = range(1, len(all_trials_df) + 1)
        
        if self.config['debug_mode']:
            total_trials = len(all_trials_df)
            print("\nTrial distribution:")
            print(f"Total trials: {total_trials}")
            
            # Count by trial type
            for trial_type in sorted(all_trials_df['trial_type'].unique()):
                type_df = all_trials_df[all_trials_df['trial_type'] == trial_type]
                print(f"Total {trial_type}: {len(type_df)}")
                
                # Count by SOA for non-baseline and non-catch trial types
                if trial_type not in ['baseline', 'catch']:
                    for soa in sorted(soa_conditions):
                        soa_count = len(type_df[type_df['soa_value_ms'] == soa])
                        print(f"  {trial_type}_{soa}: {soa_count}")
                elif trial_type == 'baseline':
                    # For baseline, show SOA distribution (though not used for looming)
                    for soa in sorted(soa_conditions):
                        soa_count = len(type_df[type_df['soa_value_ms'] == soa])
                        print(f"  baseline_{soa}: {soa_count}")
                elif trial_type == 'catch':
                    # For catch trials, show stimulus distribution
                    catch_count = len(type_df)
                    print(f"  catch_trials: {catch_count} (no tactile stimulus, no SOA)")
            
            # Show stimulus type distribution
            print("\nStimulus type distribution by trial type:")
            for trial_type in sorted(all_trials_df['trial_type'].unique()):
                if trial_type == 'baseline':
                    continue
                subset = all_trials_df[all_trials_df['trial_type'] == trial_type]
                print(f"  {trial_type}: {subset['stimulus_type'].value_counts().to_dict()}")
        
        return all_trials_df

    # ------------------------------------------------------------------
    # Assign scaffold retention slots to trials while strictly maintaining the breathphase cycle.
    # ------------------------------------------------------------------
    def assign_breath_holds(self, design_df, participant_id):
        """
        Assign scaffold retention slots to trials while strictly maintaining the breathphase cycle.
        
        Key requirements:
        1. Breathphase cycle (inhale-exhale-inhale-exhale) must remain intact
        2. Inhalation trials must go to inhale breathphase slots
        3. Exhalation trials must go to exhale breathphase slots
        4. Baseline and catch trials must be equally distributed between inhale and exhale breathphases
        5. The first N slots (where N is practice_trials) are reserved for practice trials
        """
        # Account for practice trials in required slots calculation
        practice_trials = self.config.get('practice_trials', 0)
        required_slots = len(design_df) + practice_trials
        scaffold_df = self.load_retention_slots(required_slots)
        
        # Ensure scaffold is properly sorted by timestamp/milliseconds
        scaffold_df = scaffold_df.sort_values(by='milliseconds').reset_index(drop=True)
        
        # Reserve the first N slots for practice trials
        practice_scaffold = scaffold_df.iloc[:practice_trials].copy() if practice_trials > 0 else pd.DataFrame()
        regular_scaffold = scaffold_df.iloc[practice_trials:].copy()
        
        # Create practice trials (empty placeholder trials)
        practice_trial_list = []
        if practice_trials > 0:
            for i, (_, row) in enumerate(practice_scaffold.iterrows()):
                breathphase = row['breathphase'].lower()
                practice_trial = {
                    'participant_id': participant_id,
                    'trial_number': i + 1,  # Start numbering from 1
                    'trial_type': 'practice',
                    'stimulus_type': None,
                    'soa_value_ms': None,
                    'jitter_ms': random.choice(self.config['jitter_options_ms']),  # Add jitter for realistic timing
                    'is_tactile': False
                }
                # Add all scaffold columns
                for col in row.index:
                    practice_trial[col] = row[col]
                
                practice_trial_list.append(practice_trial)
        
        # Partition design trials by trial type
        pool_inh = [d for d in design_df.to_dict('records') if d['trial_type'] == 'inhalation']
        pool_exh = [d for d in design_df.to_dict('records') if d['trial_type'] == 'exhalation']
        pool_baseline = [d for d in design_df.to_dict('records') if d['trial_type'] == 'baseline']
        pool_catch = [d for d in design_df.to_dict('records') if d['trial_type'] == 'catch']
        
        # Count scaffold slots by breathphase
        inhale_slots = regular_scaffold[regular_scaffold['breathphase'].str.lower() == 'inhale']
        exhale_slots = regular_scaffold[regular_scaffold['breathphase'].str.lower() == 'exhale']
        n_inhale_slots = len(inhale_slots)
        n_exhale_slots = len(exhale_slots)
        
        if self.config['debug_mode']:
            print(f"Scaffold: {n_inhale_slots} inhale slots, {n_exhale_slots} exhale slots")
            print(f"Trials: {len(pool_inh)} inhalation, {len(pool_exh)} exhalation, " 
                  f"{len(pool_baseline)} baseline, {len(pool_catch)} catch")
            print(f"Practice trials: {practice_trials}")
        
        # Verify we have enough slots for mandatory trials
        if len(pool_inh) > n_inhale_slots:
            raise ValueError(f"Not enough inhale slots ({n_inhale_slots}) for inhalation trials ({len(pool_inh)})")
        if len(pool_exh) > n_exhale_slots:
            raise ValueError(f"Not enough exhale slots ({n_exhale_slots}) for exhalation trials ({len(pool_exh)})")
        
        # Calculate remaining slots per breathphase
        remaining_inhale = n_inhale_slots - len(pool_inh)
        remaining_exhale = n_exhale_slots - len(pool_exh)
        
        # Total baseline+catch trials
        total_flex_trials = len(pool_baseline) + len(pool_catch)
        
        # Verify we have enough slots for all trials
        if remaining_inhale + remaining_exhale != total_flex_trials:
            raise ValueError(f"Mismatch between available slots ({remaining_inhale + remaining_exhale}) "
                            f"and flexible trials ({total_flex_trials})")
        
        # Ensure equal distribution of baseline and catch trials between breathphases
        
        # Shuffle baseline and catch trials
        random.shuffle(pool_baseline)
        random.shuffle(pool_catch)
        
        # Calculate how many of each should go to each breathphase
        baseline_inhale = len(pool_baseline) // 2
        baseline_exhale = len(pool_baseline) - baseline_inhale
        
        catch_inhale = len(pool_catch) // 2
        catch_exhale = len(pool_catch) - catch_inhale
        
        # Create inhale and exhale flexible trial pools
        inhale_flexible = pool_baseline[:baseline_inhale] + pool_catch[:catch_inhale]
        exhale_flexible = pool_baseline[baseline_inhale:] + pool_catch[catch_inhale:]
        
        random.shuffle(inhale_flexible)
        random.shuffle(exhale_flexible)
        
        if self.config['debug_mode']:
            print(f"Distributing flexible trials:")
            print(f"  Baseline: {baseline_inhale} to inhale, {baseline_exhale} to exhale")
            print(f"  Catch: {catch_inhale} to inhale, {catch_exhale} to exhale")
        
        # Create final allocation pools
        inhale_pool = pool_inh + inhale_flexible
        exhale_pool = pool_exh + exhale_flexible
        
        # Shuffle within each pool to randomize order
        random.shuffle(inhale_pool)
        random.shuffle(exhale_pool)
        
        # Create a dictionary mapping breathphase to trial pool
        phase_to_pool = {
            'inhale': inhale_pool,
            'exhale': exhale_pool
        }
        
        # Initialize counters for each breathphase
        pool_counters = {
            'inhale': 0,
            'exhale': 0
        }
        
        # Final list to store assigned trials in scaffold order
        regular_assignments = []
        
        # Assign trials to regular scaffold rows, maintaining the original order
        for _, row in regular_scaffold.iterrows():
            breathphase = row['breathphase'].lower()
            
            # Get the next trial from the appropriate pool
            if pool_counters[breathphase] < len(phase_to_pool[breathphase]):
                trial = phase_to_pool[breathphase][pool_counters[breathphase]]
                pool_counters[breathphase] += 1
                
                # Combine scaffold information with trial information
                combined = trial.copy()
                for col in row.index:
                    combined[col] = row[col]
                
                regular_assignments.append(combined)
        
        # Combine practice trials and regular trials
        final_assignments = practice_trial_list + regular_assignments
        
        # Create final dataframe and sort by scaffold order
        final_df = pd.DataFrame(final_assignments).sort_values(by='milliseconds').reset_index(drop=True)
        
        # Renumber trials from 1 to N to ensure sequential numbering
        final_df['trial_number'] = range(1, len(final_df) + 1)
        
        # Compute additional timing information
        final_df['jittered_ms'] = final_df['milliseconds'] + final_df['jitter_ms'].fillna(0)
        
        # Add looming stimulus timestamp for non-baseline trials
        looming_mask = ~final_df['trial_type'].isin(['baseline', 'practice'])
        final_df.loc[looming_mask, 'looming_stimulus_timestamp'] = final_df.loc[looming_mask, 'jittered_ms'].apply(convert_ms_to_timestamp)
        
        # Add SOA timing for non-catch, non-practice trials
        non_catch_mask = ~final_df['trial_type'].isin(['catch', 'practice'])
        final_df.loc[non_catch_mask, 'soa_ms'] = final_df.loc[non_catch_mask, 'jittered_ms'] + final_df.loc[non_catch_mask, 'soa_value_ms'].fillna(0)
        final_df.loc[non_catch_mask, 'tactile_stimulus_timestamp'] = final_df.loc[non_catch_mask, 'soa_ms'].apply(convert_ms_to_timestamp)
        
        # Verify the assignment respects the breathphase constraints
        self._validate_breathphase_assignment(final_df)
        
        return final_df

    def _validate_breathphase_assignment(self, final_df):
        """Verify that the trial assignment respects breathphase constraints."""
        if not self.config['debug_mode']:
            return
            
        print("\n=== BREATHPHASE ASSIGNMENT VALIDATION ===")
        
        # Skip practice trials for validation
        non_practice_df = final_df[final_df['trial_type'] != 'practice']
        
        # Check for inhalation trials assigned to inhale
        inh_trials = non_practice_df[non_practice_df['trial_type'] == 'inhalation']
        inh_correct = (inh_trials['breathphase'].str.lower() == 'inhale').all() if len(inh_trials) > 0 else True
        print(f"Inhalation trials on inhale breathphase: {inh_correct}")
        
        # Check for exhalation trials assigned to exhale
        exh_trials = non_practice_df[non_practice_df['trial_type'] == 'exhalation']
        exh_correct = (exh_trials['breathphase'].str.lower() == 'exhale').all() if len(exh_trials) > 0 else True
        print(f"Exhalation trials on exhale breathphase: {exh_correct}")
        
        # Count distribution of baseline and catch trials across breathphases
        baseline_inhale = ((non_practice_df['trial_type'] == 'baseline') & 
                         (non_practice_df['breathphase'].str.lower() == 'inhale')).sum()
        baseline_exhale = ((non_practice_df['trial_type'] == 'baseline') & 
                          (non_practice_df['breathphase'].str.lower() == 'exhale')).sum()
        catch_inhale = ((non_practice_df['trial_type'] == 'catch') & 
                       (non_practice_df['breathphase'].str.lower() == 'inhale')).sum()
        catch_exhale = ((non_practice_df['trial_type'] == 'catch') & 
                       (non_practice_df['breathphase'].str.lower() == 'exhale')).sum()
        
        print(f"Baseline trials: {baseline_inhale} on inhale, {baseline_exhale} on exhale")
        print(f"Catch trials: {catch_inhale} on inhale, {catch_exhale} on exhale")
        
        # Verify breathphase cycle is maintained
        prev_phase = None
        cycle_maintained = True
        expected_cycle = ['inhale', 'exhale'] * (len(final_df) // 2 + 1)
        
        for i, phase in enumerate(final_df['breathphase'].str.lower()):
            if i < len(expected_cycle) and phase != expected_cycle[i]:
                cycle_maintained = False
                print(f"Breathphase cycle disrupted at index {i}: expected {expected_cycle[i]}, got {phase}")
                break
        
        print(f"Breathphase cycle maintained: {cycle_maintained}")
        
        # Count practice trials
        practice_count = (final_df['trial_type'] == 'practice').sum()
        print(f"Practice trials: {practice_count}")

    # ------------------------------------------------------------------
    # Finalize the design CSV: Retain the scaffold column order and append new columns.
    # Also, for baseline trials remove looming stimulus info; for catch trials remove tactile stimulus info.
    # ------------------------------------------------------------------
    def finalize_design_csv(self, design_df):
        baseline_mask = (design_df['trial_type'] == 'baseline')
        design_df.loc[baseline_mask, 'stimulus_type'] = None
        design_df.loc[baseline_mask, 'looming_stimulus_timestamp'] = None
        
        catch_mask = (design_df['trial_type'] == 'catch')
        design_df.loc[catch_mask, 'tactile_stimulus_timestamp'] = None
        
        practice_mask = (design_df['trial_type'] == 'practice')
        design_df.loc[practice_mask, 'stimulus_type'] = None
        design_df.loc[practice_mask, 'looming_stimulus_timestamp'] = None
        design_df.loc[practice_mask, 'tactile_stimulus_timestamp'] = None
        
        # Set expected_response based on trial type
        design_df['expected_response'] = ~design_df['trial_type'].isin(['catch', 'practice'])
        
        scaffold_cols = self.scaffold_order if hasattr(self, 'scaffold_order') else ['timestamp', 'milliseconds', 'sample_index', 'description', 'breathphase']
        new_cols = ['participant_id', 'trial_number', 'trial_type', 'stimulus_type', 'soa_value_ms', 'jitter_ms', 'is_tactile', 'looming_stimulus_timestamp', 'tactile_stimulus_timestamp', 'soa_ms', 'jittered_ms', 'expected_response']
        final_order = scaffold_cols + [col for col in new_cols if col not in scaffold_cols and col in design_df.columns]
        remaining = [col for col in design_df.columns if col not in final_order]
        final_order.extend(remaining)
        design_df = design_df[final_order]
        
        self._validate_design(design_df)
        return design_df

    # ------------------------------------------------------------------
    # Validation: Print distributions and sequence properties.
    # ------------------------------------------------------------------
    def _validate_design(self, design_df):
        """Perform comprehensive validation and counterbalancing checks on the design."""
        if not self.config['debug_mode']:
            return
            
        print("\n=== DESIGN VALIDATION ===")
        
        # Separate practice trials from experimental trials
        practice_df = design_df[design_df['trial_type'] == 'practice']
        exp_df = design_df[design_df['trial_type'] != 'practice']
        
        # 1. Overall trial count summary
        total_trials = len(design_df)
        practice_count = len(practice_df)
        experimental_count = len(exp_df)
        
        trial_counts = exp_df['trial_type'].value_counts().to_dict()
        
        print(f"TOTAL TRIALS: {total_trials} (Practice: {practice_count}, Experimental: {experimental_count})")
        print("Trial type distribution (excluding practice):")
        for trial_type, count in sorted(trial_counts.items()):
            percent = (count / experimental_count) * 100
            print(f"  {trial_type.upper()}: {count} trials ({percent:.1f}%)")
        
        # 2. Detailed breakdown by trial type
        print("\nStimulus distribution by trial type:")
        for trial_type in sorted(exp_df['trial_type'].unique()):
            if trial_type == 'baseline':
                continue
            subset = exp_df[exp_df['trial_type'] == trial_type]
            stim_counts = subset['stimulus_type'].value_counts().to_dict()
            print(f"  {trial_type.upper()}:")
            for stim, count in sorted(stim_counts.items()):
                percent = (count / len(subset)) * 100
                print(f"    {stim}: {count} trials ({percent:.1f}%)")
        
        # 3. SOA distribution checks
        print("\nSOA distribution by trial type:")
        for trial_type in sorted(exp_df['trial_type'].unique()):
            if trial_type == 'baseline':
                continue
            subset = exp_df[exp_df['trial_type'] == trial_type]
            soa_counts = subset['soa_value_ms'].value_counts().to_dict()
            print(f"  {trial_type.upper()}:")
            for soa, count in sorted(soa_counts.items()):
                percent = (count / len(subset)) * 100
                print(f"    {soa} ms: {count} trials ({percent:.1f}%)")
        
        # 4. Jitter distribution
        print("\nJitter distribution (excluding practice):")
        jitter_counts = exp_df['jitter_ms'].value_counts().to_dict()
        for jitter, count in sorted(jitter_counts.items()):
            percent = (count / experimental_count) * 100
            print(f"  {jitter} ms: {count} trials ({percent:.1f}%)")
        
        # 5. Breath phase checks for experimental trials
        print("\nBreath phase distribution (excluding practice):")
        if 'breathphase' in exp_df.columns:
            phase_counts = exp_df['breathphase'].value_counts().to_dict()
            for phase, count in sorted(phase_counts.items()):
                percent = (count / experimental_count) * 100
                print(f"  {phase}: {count} trials ({percent:.1f}%)")
        
        # 6. Cross-tabulation of trial types by breath phase
        print("\nTrial types by breath phase (COUNTERBALANCE CHECK):")
        if 'breathphase' in exp_df.columns:
            trial_by_phase = pd.crosstab(
                exp_df['trial_type'], 
                exp_df['breathphase']
            )
            print(trial_by_phase)
            
            # Calculate chi-square statistic for independence (excluding mandatory assignments)
            try:
                # Create subset without inhalation/exhalation (which are deterministically assigned)
                flex_df = exp_df[exp_df['trial_type'].isin(['baseline', 'catch'])]
                if len(flex_df) > 0:
                    flex_table = pd.crosstab(flex_df['trial_type'], flex_df['breathphase'])
                    chi2 = 0
                    row_sums = flex_table.sum(axis=1)
                    col_sums = flex_table.sum(axis=0) 
                    total = flex_table.sum().sum()
                    
                    for i in flex_table.index:
                        for j in flex_table.columns:
                            observed = flex_table.loc[i, j]
                            expected = row_sums[i] * col_sums[j] / total
                            if expected > 0:
                                chi2 += ((observed - expected) ** 2) / expected
                    
                    print(f"\nChi-square for baseline/catch distribution: {chi2:.2f}")
                    print(f"Lower values indicate better balance between breathphases.")
            except Exception as e:
                print(f"Error calculating chi-square: {e}")
        
        # 7. Consecutive trials analysis (excluding practice)
        trial_sequence = exp_df['trial_type'].tolist()
        consecutive_counts = {}
        
        if trial_sequence:
            current_type = trial_sequence[0]
            current_count = 1
            for t in trial_sequence[1:]:
                if t == current_type:
                    current_count += 1
                else:
                    key = f"{current_type}_{current_count}"
                    consecutive_counts[key] = consecutive_counts.get(key, 0) + 1
                    current_type = t
                    current_count = 1
            key = f"{current_type}_{current_count}"
            consecutive_counts[key] = consecutive_counts.get(key, 0) + 1
            
            print("\nRuns of consecutive trial types (excluding practice):")
            for key, count in sorted(consecutive_counts.items()):
                print(f"  {key}: {count}")
        
        # 8. SOA and stimulus type correlation check (excluding practice)
        print("\nSOA-by-stimulus-type COUNTERBALANCE CHECK (excluding practice):")
        non_baseline = exp_df[exp_df['trial_type'] != 'baseline']
        if len(non_baseline) > 0:
            soa_by_stim = pd.crosstab(
                non_baseline['soa_value_ms'], 
                non_baseline['stimulus_type']
            )
            print(soa_by_stim)
            
            # Calculate coefficient of variation as a balance metric
            cv_values = []
            for _, row in soa_by_stim.iterrows():
                if row.sum() > 0:
                    cv = row.std() / row.mean() if row.mean() > 0 else 0
                    cv_values.append(cv)
            
            mean_cv = sum(cv_values) / len(cv_values) if cv_values else 0
            print(f"\nCoefficient of variation across SOA-stimulus combinations: {mean_cv:.4f}")
            print(f"Lower values indicate better counterbalancing.")
        
        # 9. Breathphase alternation check
        if 'breathphase' in design_df.columns:
            phases = design_df.sort_values('milliseconds')['breathphase'].str.lower().tolist()
            alternation_violations = 0
            for i in range(1, len(phases)):
                if phases[i] == phases[i-1]:
                    alternation_violations += 1
            
            print(f"\nBreathphase alternation violations: {alternation_violations}")
            print(f"This should be 0 - the inhale/exhale cycle must be maintained.")
        
        # 10. Timing analysis
        if 'jittered_ms' in design_df.columns:
            time_diffs = []
            sorted_df = design_df.sort_values('jittered_ms')
            for i in range(1, len(sorted_df)):
                diff = sorted_df.iloc[i]['jittered_ms'] - sorted_df.iloc[i-1]['jittered_ms']
                time_diffs.append(diff)
            
            if time_diffs:
                min_diff = min(time_diffs)
                max_diff = max(time_diffs)
                avg_diff = sum(time_diffs) / len(time_diffs)
                print(f"\nStimulus timing intervals:")
                print(f"  Minimum interval: {min_diff:.1f} ms")
                print(f"  Maximum interval: {max_diff:.1f} ms")
                print(f"  Average interval: {avg_diff:.1f} ms")
        
        # 11. Practice trials validation
        if not practice_df.empty:
            print("\nPractice trials analysis:")
            practice_phase_counts = practice_df['breathphase'].value_counts().to_dict() if 'breathphase' in practice_df.columns else {}
            print(f"  Total practice trials: {len(practice_df)}")
            for phase, count in sorted(practice_phase_counts.items()):
                print(f"  {phase}: {count} trials")
            
            print(f"  Practice trial numbers: {practice_df['trial_number'].tolist()}")
            print(f"  Practice trials have jitter: {'jitter_ms' in practice_df.columns and not practice_df['jitter_ms'].isnull().all()}")
    
    # ------------------------------------------------------------------
    # Process a single participant: generate design, assign slots, finalize, and save CSV.
    # ------------------------------------------------------------------
    def process_participant(self, participant_id):
        try:
            if self.config['debug_mode']:
                print(f"\n=== Processing participant {participant_id} ===")
            design_df = self.generate_counterbalanced_design(participant_id)
            design_df = self.assign_breath_holds(design_df, participant_id)
            design_df = self.finalize_design_csv(design_df)
            out_path = os.path.join(self.experiment_log_dir, f"participant_{participant_id}_design.csv")
            design_df.to_csv(out_path, index=False)
            if self.config['debug_mode']:
                print(f"Saved participant {participant_id} design to {out_path}")
            return True, out_path
        except Exception as e:
            print(f"Error processing participant {participant_id}: {e}")
            traceback.print_exc()
            return False, None
    
    # ------------------------------------------------------------------
    # Run design generation for multiple participants.
    # ------------------------------------------------------------------
    def run(self, participant_ids=None, num_participants=None):
        if participant_ids is None and num_participants is None:
            num_participants = self.config.get('num_participants', 5)
        if participant_ids is None:
            participant_ids = list(range(num_participants))
        if self.config['debug_mode']:
            print("=== Starting PPSDesignGenerator ===")
            print(f"Repetitions: {self.config['repetitions']}")
            print(f"Practice trials: {self.config['practice_trials']}")
            print(f"Participants: {participant_ids}")
            print(f"Box breathing settings: {self.config['box_breathing']}")
        results = {}
        for pid in participant_ids:
            success, path = self.process_participant(pid)
            results[pid] = {'success': success, 'csv_path': path}
        
        # Summary of all participants
        if self.config['debug_mode'] and len(results) > 1:
            print("\n=== CROSS-PARTICIPANT VALIDATION ===")
            all_success = all(info['success'] for info in results.values())
            print(f"All participants generated successfully: {all_success}")
        
        return results

# ----------------------------------------------------------------------
# AUDIO GENERATION PLACEHOLDER FUNCTION
# ----------------------------------------------------------------------
def generate_audio_from_design(design_csv_path, output_dir, base_audio_path=None):
    """
    Generate audio files from a design CSV.
    (Placeholder function – implement audio generation as needed.)
    """
    print(f"Audio generation from {design_csv_path} not implemented yet")
    return None, None

# ----------------------------------------------------------------------
# MAIN FUNCTION
# ----------------------------------------------------------------------
def main():
    """Main function to run the design generator."""
    # Print configuration summary
    print("\n===== PPS EXPERIMENT DESIGN GENERATOR =====")
    print(f"Repetitions (trials per SOA per condition): {CONFIG['repetitions']}")
    print(f"Practice trials at beginning: {CONFIG['practice_trials']}")
    print(f"SOA Conditions: {CONFIG['soa_conditions_ms']}")
    print(f"Jitter Options: {CONFIG['jitter_options_ms']}")
    print(f"Looming Stimuli: {LOOMING_STIMULI}")
    print(f"Number of Participants: {CONFIG['num_participants']}")
    
    # Calculate expected trial counts
    reps = CONFIG['repetitions']
    soa_count = len(CONFIG['soa_conditions_ms'])
    
    trials_per_condition = reps * soa_count
    enabled_conditions = [cond for cond, enabled in CONFIG['included_conditions'].items() if enabled]
    total_base_trials = trials_per_condition * len(enabled_conditions)
    catch_trials = int(np.ceil(total_base_trials * CONFIG['catch_trial_percentage'] / 100))
    practice_trials = CONFIG['practice_trials']
    
    print("\n===== EXPECTED TRIAL COUNTS =====")
    print(f"Trials per condition: {reps} trials per SOA × {soa_count} SOAs = {trials_per_condition}")
    for cond in enabled_conditions:
        print(f"  {cond.upper()}: {trials_per_condition} trials")
    print(f"Catch trials ({CONFIG['catch_trial_percentage']}%): {catch_trials}")
    print(f"Practice trials: {practice_trials}")
    print(f"TOTAL TRIALS PER PARTICIPANT: {total_base_trials + catch_trials + practice_trials}")
    
    # Run the generator
    generator = PPSDesignGenerator(CONFIG)
    results = generator.run()
    
    # Summarize results
    print("\n===== GENERATION SUMMARY =====")
    success_count = sum(1 for info in results.values() if info['success'])
    print(f"Successfully generated designs for {success_count}/{len(results)} participants.")
    
    for pid, info in results.items():
        if info['success']:
            print(f"Participant {pid}: CSV generated at {info['csv_path']}")
        else:
            print(f"Participant {pid}: ERROR - no CSV generated.")
    
    print("\nTo generate audio files from these designs, implement the audio generation function.")

if __name__ == "__main__":
    main()
```

```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-
"""
PPS Design Generator with WAV Marker Integration

This script generates participant-specific experimental design CSVs for a 
Peripersonal Space (PPS) experiment. It uses a WAV marker-based approach to:

1. Extract breath markers directly from WAV files as the scaffold
2. Match the exact format from the WAV marker extraction script
3. Assign trials to these markers while maintaining the breathphase cycle
4. Generate participant-specific design files with all required trial parameters

Key features:
- Extracts WAV markers directly from audio files using odd/even ID encoding
- Preserves the exact column structure from the marker extraction
- Maintains all counterbalancing and randomization functionality
- Ensures proper assignment of trial types to breath phases

Author: AI Assistant
"""

import os
import numpy as np
import pandas as pd
import random
import traceback
import wave
import struct
from pathlib import Path
import io

# ----------------------------------------------------------------------
# CONFIGURATION PARAMETERS
# ----------------------------------------------------------------------
CONFIG = {
    'repetitions': 12,           # Number of trials per SOA condition
    'num_participants': 100,     # Default number of participants
    'practice_trials': 8,        # Number of practice trials at the beginning
    'included_conditions': {
        'inhalation': True,
        'exhalation': True,
        'baseline': True,
    },
    'catch_trial_percentage': 10,
    'soa_conditions_ms': [190, 400, 700, 1000, 1500],
    'jitter_options_ms': [100, 200, 300, 400, 500],
    'debug_mode': True,
    'paths': {
        # Directory where participant-specific CSV files will be saved
        'experiment_log_dir': r"C:\Users\cogpsy-vrlab\Documents\GitHub\BreathingSpace\Level1_AudioGeneration\ExperimentLog",
        # Directory containing marker-based audio files and accompanying CSV
        'experiment_audio_dir': r"C:\Users\cogpsy-vrlab\Documents\GitHub\BreathingSpace\Level1_AudioGeneration\ExperimentAudio",
        # Base name for parent WAV file with embedded markers to use as scaffold (should be in experiment_audio_dir)
        'parent_wav_name': "BoxBreathing_complete.wav",
    },
    'box_breathing': {
        'cycle_duration_sec': 8,      # Duration of one breath cycle in seconds
        'sample_rate': 48000,         # Audio sample rate (samples per second)
    }
}

# Noise types for looming stimuli (can be any labels)
LOOMING_STIMULI = ['blue', 'brown', 'pink', 'white']

# Marker type constants for WAV file marker extraction
MARKER_TYPE_INHALATION = 1  # Odd IDs: 1, 3, 5, 7, ...
MARKER_TYPE_EXHALATION = 2  # Even IDs: 2, 4, 6, 8, ...

# ----------------------------------------------------------------------
# Helper function: Convert milliseconds to "MM:SS.s" format
# ----------------------------------------------------------------------
def convert_ms_to_timestamp(ms):
    """
    Convert milliseconds to timestamp in MM:SS.s format.
    """
    if ms is None:
        return None
        
    m = int(ms // 60000)
    s = (ms % 60000) / 1000.0
    return f"{m:02}:{s:.1f}"

# ----------------------------------------------------------------------
# Function to extract markers directly from a WAV file
# ----------------------------------------------------------------------
def extract_markers_from_wav(wav_file):
    """
    Extract marker information directly from a WAV file's cue points 
    using odd/even ID detection.
    
    Args:
        wav_file: Path to WAV file with embedded markers
        
    Returns:
        List of dictionaries with marker information
    """
    # Try to read sample rate from WAV file
    sample_rate = 48000  # Default fallback
    try:
        with wave.open(wav_file, 'rb') as wf:
            sample_rate = wf.getframerate()
    except Exception as e:
        print(f"Warning: Couldn't read sample rate from WAV file: {e}")
        print(f"Using default sample rate: {sample_rate} Hz")
    
    markers = []
    
    try:
        with open(wav_file, 'rb') as f:
            # Check RIFF header
            if f.read(4) != b'RIFF':
                raise ValueError("Not a valid RIFF file")
            
            # Read file size (and skip)
            f.read(4)
            
            # Check WAVE format
            if f.read(4) != b'WAVE':
                raise ValueError("Not a valid WAVE file")
            
            # Find cue chunk
            while True:
                chunk_pos = f.tell()
                chunk_header = f.read(4)
                
                if not chunk_header or len(chunk_header) < 4:
                    break  # End of file
                
                chunk_size = int.from_bytes(f.read(4), byteorder='little')
                
                if chunk_header == b'cue ':
                    # Found cue chunk
                    num_cue_points = int.from_bytes(f.read(4), byteorder='little')
                    print(f"Found {num_cue_points} cue points in WAV file")
                    
                    # Read all cue points
                    for i in range(num_cue_points):
                        # Read cue point data
                        cue_id = int.from_bytes(f.read(4), byteorder='little')
                        position = int.from_bytes(f.read(4), byteorder='little')
                        data_chunk_id = f.read(4)  # Should be 'data'
                        chunk_start = int.from_bytes(f.read(4), byteorder='little')
                        block_start = int.from_bytes(f.read(4), byteorder='little')
                        sample_offset = int.from_bytes(f.read(4), byteorder='little')
                        
                        # Determine marker type from cue ID parity (odd/even)
                        marker_type = None
                        breathphase = None
                        
                        if cue_id % 2 == 1:  # Odd ID = Inhale
                            marker_type = MARKER_TYPE_INHALATION
                            breathphase = "inhale"
                        elif cue_id % 2 == 0:  # Even ID = Exhale
                            marker_type = MARKER_TYPE_EXHALATION
                            breathphase = "exhale"
                        else:
                            # This condition can't actually happen (every number is either odd or even)
                            continue
                        
                        # Calculate timestamp
                        timestamp_seconds = position / sample_rate
                        milliseconds = int(timestamp_seconds * 1000)
                        
                        # Format timestamps
                        minutes = int(timestamp_seconds // 60)
                        seconds = timestamp_seconds % 60
                        retentionslot_timestamp = f"{minutes:02}:{seconds:.1f}"
                        
                        # Create marker entry with standard column format
                        marker = {
                            'retentionslot_timestamp': retentionslot_timestamp,
                            'milliseconds': milliseconds,
                            'sample_index': position,
                            'description': '',  # Empty as in original scaffold
                            'breathphase': breathphase,
                            'marker_type': marker_type
                        }
                        
                        markers.append(marker)
                    
                    break  # Found what we needed
                else:
                    # Skip this chunk
                    f.seek(chunk_pos + 8 + chunk_size)
        
    except Exception as e:
        print(f"Error reading markers from WAV file: {e}")
        import traceback
        traceback.print_exc()
    
    # Sort markers by timestamp
    markers.sort(key=lambda x: x['milliseconds'])
    
    if not markers:
        print("No markers found in WAV file")
    else:
        print(f"Extracted {len(markers)} markers from WAV file")
        print(f"Inhale markers: {sum(1 for m in markers if m['breathphase'] == 'inhale')}")
        print(f"Exhale markers: {sum(1 for m in markers if m['breathphase'] == 'exhale')}")
    
    return markers

# ----------------------------------------------------------------------
# CLASS: PPSDesignGenerator
# ----------------------------------------------------------------------
class PPSDesignGenerator:
    """Generates participant-specific CSV design files for PPS experiments."""
    
    def __init__(self, config=None):
        self.config = config or CONFIG
        
        # Prepare output directory
        self.experiment_log_dir = self.config['paths']['experiment_log_dir']
        os.makedirs(self.experiment_log_dir, exist_ok=True)
        
        # Calculate trial counts
        self.trial_counts = self._calculate_trial_counts()
        self.conditions = [cond for cond, inc in self.config['included_conditions'].items() if inc]
        
        # Set seed for reproducibility
        self.base_seed = 42
        
        # Look for parent WAV file in the experiment_audio_dir
        self.experiment_audio_dir = self.config['paths']['experiment_audio_dir']
        self.parent_wav_name = self.config['paths']['parent_wav_name']
        self.marker_wav_file = os.path.join(self.experiment_audio_dir, self.parent_wav_name)
        
        # Also look for accompanying CSV with the same base name
        self.marker_csv_file = os.path.join(
            self.experiment_audio_dir, 
            self.parent_wav_name.replace('.wav', '_markers.csv')
        )
        
        # First try to read markers from WAV file
        self.wav_markers = []
        if os.path.exists(self.marker_wav_file):
            print(f"Found parent WAV file: {self.marker_wav_file}")
            self.wav_markers = extract_markers_from_wav(self.marker_wav_file)
        else:
            print(f"Warning: Parent WAV file not found at: {self.marker_wav_file}")
        
        # If no markers from WAV or too few, try reading from CSV
        if not self.wav_markers or len(self.wav_markers) < 10:
            print(f"Looking for marker CSV file: {self.marker_csv_file}")
            if os.path.exists(self.marker_csv_file):
                try:
                    csv_df = pd.read_csv(self.marker_csv_file)
                    print(f"Found marker CSV with {len(csv_df)} rows")
                    
                    # Convert CSV to marker format
                    self.wav_markers = []
                    for _, row in csv_df.iterrows():
                        marker = {}
                        for col in row.index:
                            marker[col] = row[col]
                        
                        # Ensure we have breathphase information
                        if 'breathphase' not in marker and 'breathPhase' in marker:
                            marker['breathphase'] = marker['breathPhase']
                        
                        # Add marker type based on breathphase
                        if 'breathphase' in marker:
                            if 'inhale' in str(marker['breathphase']).lower():
                                marker['marker_type'] = MARKER_TYPE_INHALATION
                            elif 'exhale' in str(marker['breathphase']).lower():
                                marker['marker_type'] = MARKER_TYPE_EXHALATION
                        
                        self.wav_markers.append(marker)
                    
                    print(f"Loaded {len(self.wav_markers)} markers from CSV")
                except Exception as e:
                    print(f"Error reading marker CSV: {e}")
            else:
                print(f"Warning: No marker CSV file found at: {self.marker_csv_file}")
                print("Will create artificial scaffold.")
        
        # Default column structure that must be preserved
        self.scaffold_columns = [
            'retentionslot_timestamp', 'milliseconds', 'sample_index', 
            'description', 'breathphase'
        ]
        
        if self.config['debug_mode']:
            total_trials = sum(self.trial_counts.values())
            print("Initialized PPSDesignGenerator with:")
            print(f"  - Repetitions: {self.config['repetitions']}")
            print(f"  - Practice trials: {self.config['practice_trials']}")
            print(f"  - Conditions: {self.conditions}")
            print(f"  - SOA values: {self.config['soa_conditions_ms']}")
            print(f"  - Total trials per participant: {total_trials} + {self.config['practice_trials']} practice = {total_trials + self.config['practice_trials']}")
            print(f"  - Output directory: {self.experiment_log_dir}")
    
    def _calculate_trial_counts(self):
        """
        Calculate the number of trials for each condition, using the repetitions parameter
        to determine how many trials per SOA per condition.
        """
        soa_conditions = self.config['soa_conditions_ms']
        repetitions = self.config['repetitions']
        catch_percentage = self.config['catch_trial_percentage']
        
        trial_counts = {}
        for condition, include in self.config['included_conditions'].items():
            if include:
                # Each condition gets repetitions × number of SOAs trials
                trial_counts[condition] = len(soa_conditions) * repetitions
        
        regular_trials = sum(trial_counts.values())
        catch_trials = int(np.ceil(regular_trials * catch_percentage / 100))
        trial_counts['catch'] = catch_trials
        
        if self.config['debug_mode']:
            print("Trial count calculation:")
            for cond, count in trial_counts.items():
                print(f"  {cond}: {count} trials")
            print(f"  Total: {sum(trial_counts.values())} trials")
        
        return trial_counts

    # ------------------------------------------------------------------
    # Create scaffold from WAV markers or fallback to CSV
    # ------------------------------------------------------------------
    def create_retention_slot_scaffold(self, required_slots):
        """
        Create a scaffold DataFrame from WAV markers or fallback to CSV.
        
        Args:
            required_slots: Minimum number of slots needed
            
        Returns:
            DataFrame with scaffold structure
        """
        if self.wav_markers and len(self.wav_markers) > 0:
            # Use extracted WAV markers as scaffold
            scaffold_df = pd.DataFrame(self.wav_markers)
            
            # Ensure all required columns exist
            for col in self.scaffold_columns:
                if col not in scaffold_df.columns:
                    scaffold_df[col] = None
                    print(f"Warning: Added missing column '{col}' to marker scaffold")
            
            print(f"Created scaffold from WAV markers: {len(scaffold_df)} slots")
            
        else:
            # Fallback: Create scaffold from scratch
            print("No WAV markers available. Creating artificial scaffold...")
            cycle_sec = self.config['box_breathing']['cycle_duration_sec']
            sample_rate = self.config['box_breathing']['sample_rate']
            
            scaffold_data = []
            current_ms = 0
            
            for i in range(required_slots):
                # Alternate breathphase
                breathphase = "inhale" if i % 2 == 0 else "exhale"
                
                # Convert to timestamp format
                timestamp = convert_ms_to_timestamp(current_ms)
                
                # Create marker entry
                marker = {
                    'retentionslot_timestamp': timestamp,
                    'milliseconds': current_ms,
                    'sample_index': int(current_ms * sample_rate / 1000),
                    'description': f"Artificial {breathphase} marker {i+1}",
                    'breathphase': breathphase,
                    'marker_type': MARKER_TYPE_INHALATION if breathphase == "inhale" else MARKER_TYPE_EXHALATION
                }
                
                scaffold_data.append(marker)
                current_ms += cycle_sec * 1000  # Add 8 seconds for next marker
            
            scaffold_df = pd.DataFrame(scaffold_data)
            print(f"Created artificial scaffold with {len(scaffold_df)} slots")
        
        # Ensure we have enough slots
        if len(scaffold_df) < required_slots:
            # Need to extend scaffold
            print(f"Extending scaffold from {len(scaffold_df)} to {required_slots} slots")
            
            last_row = scaffold_df.iloc[-1]
            last_ms = float(last_row['milliseconds'])
            last_sample_index = int(last_row['sample_index'])
            cycle_sec = self.config['box_breathing']['cycle_duration_sec']
            sample_rate = self.config['box_breathing']['sample_rate']
            
            # Get the last breathphase to ensure proper alternation
            last_breathphase = last_row['breathphase'].lower()
            
            # Create additional rows
            new_rows = []
            for i in range(required_slots - len(scaffold_df)):
                # Increment by cycle duration for each new row
                new_ms = last_ms + (cycle_sec * 1000) * (i + 1)
                new_sample_index = last_sample_index + int(cycle_sec * sample_rate * (i + 1))
                new_timestamp = convert_ms_to_timestamp(new_ms)
                
                # Alternate breathphase
                if i == 0:  # First new row
                    new_breathphase = 'inhale' if last_breathphase == 'exhale' else 'exhale'
                else:  # Subsequent rows
                    prev_phase = new_rows[i-1]['breathphase'].lower()
                    new_breathphase = 'inhale' if prev_phase == 'exhale' else 'exhale'
                
                # Create marker entry
                new_row = {
                    'retentionslot_timestamp': new_timestamp,
                    'milliseconds': new_ms,
                    'sample_index': new_sample_index,
                    'description': f"Extended {new_breathphase} marker",
                    'breathphase': new_breathphase,
                    'marker_type': MARKER_TYPE_INHALATION if new_breathphase == "inhale" else MARKER_TYPE_EXHALATION
                }
                
                new_rows.append(new_row)
            
            # Append new rows to scaffold
            if new_rows:
                scaffold_df = pd.concat([scaffold_df, pd.DataFrame(new_rows)], ignore_index=True)
        
        # Trim if we have too many slots
        if len(scaffold_df) > required_slots:
            scaffold_df = scaffold_df.iloc[:required_slots]
        
        # Ensure scaffold is sorted by time
        scaffold_df = scaffold_df.sort_values(by='milliseconds').reset_index(drop=True)
        
        return scaffold_df

    # ------------------------------------------------------------------
    # Generate counterbalanced design (unchanged)
    # ------------------------------------------------------------------
    def generate_counterbalanced_design(self, participant_id):
        """
        Generate a counterbalanced design for a participant with trials per SOA
        determined by the repetitions parameter in the config.
        
        Catch trials don't have SOA values assigned since they don't have tactile stimuli.
        """
        random.seed(self.base_seed + participant_id)
        np.random.seed(self.base_seed + participant_id)
        
        soa_conditions = self.config['soa_conditions_ms']
        jitter_options = self.config['jitter_options_ms']
        stimulus_types = LOOMING_STIMULI
        repetitions = self.config['repetitions']  # Trials per SOA per condition
        
        all_trials = []
        
        # Generate main conditions: inhalation, exhalation, and baseline
        for condition, count in self.trial_counts.items():
            if condition == 'catch' or count == 0:
                continue
            
            # For each SOA, generate exactly 'repetitions' trials
            for soa in soa_conditions:
                # Calculate how many trials per stimulus type (as evenly as possible)
                trials_per_soa = repetitions
                trials_per_stim = trials_per_soa // len(stimulus_types)  # Integer division
                remainder = trials_per_soa % len(stimulus_types)
                
                # Distribute trials across stimulus types
                stim_trials = []
                for i, stim_type in enumerate(stimulus_types):
                    # Add extra trial for remainder distribution
                    extra = 1 if i < remainder else 0
                    for _ in range(trials_per_stim + extra):
                        stim_trials.append(stim_type)
                
                # Shuffle the stimulus types
                random.shuffle(stim_trials)
                
                # Create trials for this SOA
                for stim_type in stim_trials:
                    all_trials.append({
                        'participant_id': participant_id,
                        'trial_type': condition,
                        'stimulus_type': stim_type,
                        'soa_value_ms': soa,
                        'jitter_ms': random.choice(jitter_options),
                        'is_tactile': True  # For these main trials, tactile is present.
                    })
        
        # Add catch trials (looming only, no tactile) - no SOA values needed
        catch_count = self.trial_counts.get('catch', 0)
        if catch_count > 0:
            # Distribute catch trials evenly across stimulus types only
            catch_trials = []
            stim_counts = {stim: 0 for stim in stimulus_types}
            
            for _ in range(catch_count):
                # Choose the stimulus type with the lowest count
                min_stim = min(stim_counts, key=stim_counts.get)
                
                catch_trials.append({
                    'participant_id': participant_id,
                    'trial_type': 'catch',
                    'stimulus_type': min_stim,
                    'soa_value_ms': None,  # No SOA for catch trials
                    'jitter_ms': random.choice(jitter_options),
                    'is_tactile': False
                })
                
                stim_counts[min_stim] += 1
            
            # Shuffle catch trials and add to all trials
            random.shuffle(catch_trials)
            all_trials.extend(catch_trials)
        
        # Shuffle all trials and assign trial numbers
        all_trials_df = pd.DataFrame(all_trials)
        all_trials_df = all_trials_df.sample(frac=1, random_state=participant_id*100).reset_index(drop=True)
        all_trials_df['trial_number'] = range(1, len(all_trials_df) + 1)
        
        if self.config['debug_mode']:
            total_trials = len(all_trials_df)
            print("\nTrial distribution:")
            print(f"Total trials: {total_trials}")
            
            # Count by trial type
            for trial_type in sorted(all_trials_df['trial_type'].unique()):
                type_df = all_trials_df[all_trials_df['trial_type'] == trial_type]
                print(f"Total {trial_type}: {len(type_df)}")
                
                # Count by SOA for non-baseline and non-catch trial types
                if trial_type not in ['baseline', 'catch']:
                    for soa in sorted(soa_conditions):
                        soa_count = len(type_df[type_df['soa_value_ms'] == soa])
                        print(f"  {trial_type}_{soa}: {soa_count}")
                elif trial_type == 'baseline':
                    # For baseline, show SOA distribution (though not used for looming)
                    for soa in sorted(soa_conditions):
                        soa_count = len(type_df[type_df['soa_value_ms'] == soa])
                        print(f"  baseline_{soa}: {soa_count}")
                elif trial_type == 'catch':
                    # For catch trials, show stimulus distribution
                    catch_count = len(type_df)
                    print(f"  catch_trials: {catch_count} (no tactile stimulus, no SOA)")
            
            # Show stimulus type distribution
            print("\nStimulus type distribution by trial type:")
            for trial_type in sorted(all_trials_df['trial_type'].unique()):
                if trial_type == 'baseline':
                    continue
                subset = all_trials_df[all_trials_df['trial_type'] == trial_type]
                print(f"  {trial_type}: {subset['stimulus_type'].value_counts().to_dict()}")
        
        return all_trials_df

    # ------------------------------------------------------------------
    # Assign scaffold retention slots to trials (modified to use markers)
    # ------------------------------------------------------------------
    def assign_breath_holds(self, design_df, participant_id):
        """
        Assign scaffold retention slots to trials while strictly maintaining the breathphase cycle.
        
        Key requirements:
        1. Breathphase cycle (inhale-exhale-inhale-exhale) must remain intact
        2. Inhalation trials must go to inhale breathphase slots
        3. Exhalation trials must go to exhale breathphase slots
        4. Baseline and catch trials must be equally distributed between inhale and exhale breathphases
        5. The first N slots (where N is practice_trials) are reserved for practice trials
        """
        # Account for practice trials in required slots calculation
        practice_trials = self.config.get('practice_trials', 0)
        required_slots = len(design_df) + practice_trials
        
        # Create or load scaffold with exactly the marker-based format
        scaffold_df = self.create_retention_slot_scaffold(required_slots)
        
        # Reserve the first N slots for practice trials
        practice_scaffold = scaffold_df.iloc[:practice_trials].copy() if practice_trials > 0 else pd.DataFrame()
        regular_scaffold = scaffold_df.iloc[practice_trials:].copy()
        
        # Create practice trials (empty placeholder trials)
        practice_trial_list = []
        if practice_trials > 0:
            for i, (_, row) in enumerate(practice_scaffold.iterrows()):
                breathphase = row['breathphase'].lower()
                practice_trial = {
                    'participant_id': participant_id,
                    'trial_number': i + 1,  # Start numbering from 1
                    'trial_type': 'practice',
                    'stimulus_type': None,
                    'soa_value_ms': None,
                    'jitter_ms': random.choice(self.config['jitter_options_ms']),
                    'is_tactile': False
                }
                # Add all scaffold columns
                for col in row.index:
                    practice_trial[col] = row[col]
                
                practice_trial_list.append(practice_trial)
        
        # Partition design trials by trial type
        pool_inh = [d for d in design_df.to_dict('records') if d['trial_type'] == 'inhalation']
        pool_exh = [d for d in design_df.to_dict('records') if d['trial_type'] == 'exhalation']
        pool_baseline = [d for d in design_df.to_dict('records') if d['trial_type'] == 'baseline']
        pool_catch = [d for d in design_df.to_dict('records') if d['trial_type'] == 'catch']
        
        # Count scaffold slots by breathphase
        inhale_slots = regular_scaffold[regular_scaffold['breathphase'].str.lower() == 'inhale']
        exhale_slots = regular_scaffold[regular_scaffold['breathphase'].str.lower() == 'exhale']
        n_inhale_slots = len(inhale_slots)
        n_exhale_slots = len(exhale_slots)
        
        if self.config['debug_mode']:
            print(f"Scaffold: {n_inhale_slots} inhale slots, {n_exhale_slots} exhale slots")
            print(f"Trials: {len(pool_inh)} inhalation, {len(pool_exh)} exhalation, " 
                  f"{len(pool_baseline)} baseline, {len(pool_catch)} catch")
            print(f"Practice trials: {practice_trials}")
        
        # Verify we have enough slots for mandatory trials
        if len(pool_inh) > n_inhale_slots:
            raise ValueError(f"Not enough inhale slots ({n_inhale_slots}) for inhalation trials ({len(pool_inh)})")
        if len(pool_exh) > n_exhale_slots:
            raise ValueError(f"Not enough exhale slots ({n_exhale_slots}) for exhalation trials ({len(pool_exh)})")
        
        # Calculate remaining slots per breathphase
        remaining_inhale = n_inhale_slots - len(pool_inh)
        remaining_exhale = n_exhale_slots - len(pool_exh)
        
        # Total baseline+catch trials
        total_flex_trials = len(pool_baseline) + len(pool_catch)
        
        # Verify we have enough slots for all trials
        if remaining_inhale + remaining_exhale != total_flex_trials:
            raise ValueError(f"Mismatch between available slots ({remaining_inhale + remaining_exhale}) "
                            f"and flexible trials ({total_flex_trials})")
        
        # Ensure equal distribution of baseline and catch trials between breathphases
        
        # Shuffle baseline and catch trials
        random.shuffle(pool_baseline)
        random.shuffle(pool_catch)
        
        # Calculate how many of each should go to each breathphase
        baseline_inhale = len(pool_baseline) // 2
        baseline_exhale = len(pool_baseline) - baseline_inhale
        
        catch_inhale = len(pool_catch) // 2
        catch_exhale = len(pool_catch) - catch_inhale
        
        # Create inhale and exhale flexible trial pools
        inhale_flexible = pool_baseline[:baseline_inhale] + pool_catch[:catch_inhale]
        exhale_flexible = pool_baseline[baseline_inhale:] + pool_catch[catch_inhale:]
        
        random.shuffle(inhale_flexible)
        random.shuffle(exhale_flexible)
        
        if self.config['debug_mode']:
            print(f"Distributing flexible trials:")
            print(f"  Baseline: {baseline_inhale} to inhale, {baseline_exhale} to exhale")
            print(f"  Catch: {catch_inhale} to inhale, {catch_exhale} to exhale")
        
        # Create final allocation pools
        inhale_pool = pool_inh + inhale_flexible
        exhale_pool = pool_exh + exhale_flexible
        
        # Shuffle within each pool to randomize order
        random.shuffle(inhale_pool)
        random.shuffle(exhale_pool)
        
        # Create a dictionary mapping breathphase to trial pool
        phase_to_pool = {
            'inhale': inhale_pool,
            'exhale': exhale_pool
        }
        
        # Initialize counters for each breathphase
        pool_counters = {
            'inhale': 0,
            'exhale': 0
        }
        
        # Final list to store assigned trials in scaffold order
        regular_assignments = []
        
        # Assign trials to regular scaffold rows, maintaining the original order
        for _, row in regular_scaffold.iterrows():
            breathphase = row['breathphase'].lower()
            
            # Get the next trial from the appropriate pool
            if pool_counters[breathphase] < len(phase_to_pool[breathphase]):
                trial = phase_to_pool[breathphase][pool_counters[breathphase]]
                pool_counters[breathphase] += 1
                
                # Combine scaffold information with trial information
                combined = trial.copy()
                for col in row.index:
                    combined[col] = row[col]
                
                regular_assignments.append(combined)
        
        # Combine practice trials and regular trials
        final_assignments = practice_trial_list + regular_assignments
        
        # Create final dataframe and sort by scaffold order
        final_df = pd.DataFrame(final_assignments).sort_values(by='milliseconds').reset_index(drop=True)
        
        # Renumber trials from 1 to N to ensure sequential numbering
        final_df['trial_number'] = range(1, len(final_df) + 1)
        
        # Compute additional timing information
        final_df['jittered_ms'] = final_df['milliseconds'] + final_df['jitter_ms'].fillna(0)
        
        # Add looming stimulus timestamp for non-baseline trials
        looming_mask = ~final_df['trial_type'].isin(['baseline', 'practice'])
        final_df.loc[looming_mask, 'looming_stimulus_timestamp'] = final_df.loc[looming_mask, 'jittered_ms'].apply(convert_ms_to_timestamp)
        
        # Add SOA timing for non-catch, non-practice trials
        non_catch_mask = ~final_df['trial_type'].isin(['catch', 'practice'])
        final_df.loc[non_catch_mask, 'soa_ms'] = final_df.loc[non_catch_mask, 'jittered_ms'] + final_df.loc[non_catch_mask, 'soa_value_ms'].fillna(0)
        final_df.loc[non_catch_mask, 'tactile_stimulus_timestamp'] = final_df.loc[non_catch_mask, 'soa_ms'].apply(convert_ms_to_timestamp)
        
        # Verify the assignment respects the breathphase constraints
        self._validate_breathphase_assignment(final_df)
        
        return final_df

    def _validate_breathphase_assignment(self, final_df):
        """Verify that the trial assignment respects breathphase constraints."""
        if not self.config['debug_mode']:
            return
            
        print("\n=== BREATHPHASE ASSIGNMENT VALIDATION ===")
        
        # Skip practice trials for validation
        non_practice_df = final_df[final_df['trial_type'] != 'practice']
        
        # Check for inhalation trials assigned to inhale
        inh_trials = non_practice_df[non_practice_df['trial_type'] == 'inhalation']
        inh_correct = (inh_trials['breathphase'].str.lower() == 'inhale').all() if len(inh_trials) > 0 else True
        print(f"Inhalation trials on inhale breathphase: {inh_correct}")
        
        # Check for exhalation trials assigned to exhale
        exh_trials = non_practice_df[non_practice_df['trial_type'] == 'exhalation']
        exh_correct = (exh_trials['breathphase'].str.lower() == 'exhale').all() if len(exh_trials) > 0 else True
        print(f"Exhalation trials on exhale breathphase: {exh_correct}")
        
        # Count distribution of baseline and catch trials across breathphases
        baseline_inhale = ((non_practice_df['trial_type'] == 'baseline') & 
                         (non_practice_df['breathphase'].str.lower() == 'inhale')).sum()
        baseline_exhale = ((non_practice_df['trial_type'] == 'baseline') & 
                          (non_practice_df['breathphase'].str.lower() == 'exhale')).sum()
        catch_inhale = ((non_practice_df['trial_type'] == 'catch') & 
                       (non_practice_df['breathphase'].str.lower() == 'inhale')).sum()
        catch_exhale = ((non_practice_df['trial_type'] == 'catch') & 
                       (non_practice_df['breathphase'].str.lower() == 'exhale')).sum()
        
        print(f"Baseline trials: {baseline_inhale} on inhale, {baseline_exhale} on exhale")
        print(f"Catch trials: {catch_inhale} on inhale, {catch_exhale} on exhale")
        
        # Verify breathphase cycle is maintained
        prev_phase = None
        cycle_maintained = True
        expected_cycle = ['inhale', 'exhale'] * (len(final_df) // 2 + 1)
        
        for i, phase in enumerate(final_df['breathphase'].str.lower()):
            if i < len(expected_cycle) and phase != expected_cycle[i]:
                cycle_maintained = False
                print(f"Breathphase cycle disrupted at index {i}: expected {expected_cycle[i]}, got {phase}")
                break
        
        print(f"Breathphase cycle maintained: {cycle_maintained}")
        
        # Count practice trials
        practice_count = (final_df['trial_type'] == 'practice').sum()
        print(f"Practice trials: {practice_count}")

    # ------------------------------------------------------------------
    # Finalize the design CSV (modified to match WAV marker format)
    # ------------------------------------------------------------------
    def finalize_design_csv(self, design_df):
        """
        Finalize the design CSV by cleaning up trial assignments and ensuring
        the WAV marker format is preserved.
        """
        # Process trial types
        baseline_mask = (design_df['trial_type'] == 'baseline')
        design_df.loc[baseline_mask, 'stimulus_type'] = None
        design_df.loc[baseline_mask, 'looming_stimulus_timestamp'] = None
        
        catch_mask = (design_df['trial_type'] == 'catch')
        design_df.loc[catch_mask, 'tactile_stimulus_timestamp'] = None
        
        practice_mask = (design_df['trial_type'] == 'practice')
        design_df.loc[practice_mask, 'stimulus_type'] = None
        design_df.loc[practice_mask, 'looming_stimulus_timestamp'] = None
        design_df.loc[practice_mask, 'tactile_stimulus_timestamp'] = None
        
        # Set expected_response based on trial type
        design_df['expected_response'] = ~design_df['trial_type'].isin(['catch', 'practice'])
        
        # Ensure we maintain the marker-based column order
        # Start with scaffold columns, then add trial-specific columns
        marker_columns = self.scaffold_columns
        trial_columns = [
            'participant_id', 'trial_number', 'trial_type', 'stimulus_type', 
            'soa_value_ms', 'jitter_ms', 'is_tactile', 'looming_stimulus_timestamp', 
            'tactile_stimulus_timestamp', 'soa_ms', 'jittered_ms', 'expected_response'
        ]
        
        # Create final column order
        all_columns = marker_columns.copy()
        for col in trial_columns:
            if col in design_df.columns and col not in all_columns:
                all_columns.append(col)
                
        # Get any remaining columns not in the predefined lists
        remaining_columns = [col for col in design_df.columns if col not in all_columns]
        all_columns.extend(remaining_columns)
        
        # Reorder and return
        final_df = design_df[all_columns].copy()
        
        self._validate_design(final_df)
        return final_df

    # ------------------------------------------------------------------
    # Design validation (unchanged)
    # ------------------------------------------------------------------
    def _validate_design(self, design_df):
        """Perform comprehensive validation and counterbalancing checks on the design."""
        if not self.config['debug_mode']:
            return
            
        print("\n=== DESIGN VALIDATION ===")
        
        # Separate practice trials from experimental trials
        practice_df = design_df[design_df['trial_type'] == 'practice']
        exp_df = design_df[design_df['trial_type'] != 'practice']
        
        # 1. Overall trial count summary
        total_trials = len(design_df)
        practice_count = len(practice_df)
        experimental_count = len(exp_df)
        
        trial_counts = exp_df['trial_type'].value_counts().to_dict()
        
        print(f"TOTAL TRIALS: {total_trials} (Practice: {practice_count}, Experimental: {experimental_count})")
        print("Trial type distribution (excluding practice):")
        for trial_type, count in sorted(trial_counts.items()):
            percent = (count / experimental_count) * 100
            print(f"  {trial_type.upper()}: {count} trials ({percent:.1f}%)")
        
        # 2. Detailed breakdown by trial type
        print("\nStimulus distribution by trial type:")
        for trial_type in sorted(exp_df['trial_type'].unique()):
            if trial_type == 'baseline':
                continue
            subset = exp_df[exp_df['trial_type'] == trial_type]
            stim_counts = subset['stimulus_type'].value_counts().to_dict()
            print(f"  {trial_type.upper()}:")
            for stim, count in sorted(stim_counts.items()):
                percent = (count / len(subset)) * 100
                print(f"    {stim}: {count} trials ({percent:.1f}%)")
        
        # 3. SOA distribution checks
        print("\nSOA distribution by trial type:")
        for trial_type in sorted(exp_df['trial_type'].unique()):
            if trial_type == 'baseline':
                continue
            subset = exp_df[exp_df['trial_type'] == trial_type]
            soa_counts = subset['soa_value_ms'].value_counts().to_dict()
            print(f"  {trial_type.upper()}:")
            for soa, count in sorted(soa_counts.items()):
                percent = (count / len(subset)) * 100
                print(f"    {soa} ms: {count} trials ({percent:.1f}%)")
        
        # 4. Jitter distribution
        print("\nJitter distribution (excluding practice):")
        jitter_counts = exp_df['jitter_ms'].value_counts().to_dict()
        for jitter, count in sorted(jitter_counts.items()):
            percent = (count / experimental_count) * 100
            print(f"  {jitter} ms: {count} trials ({percent:.1f}%)")
        
        # 5. Breath phase checks for experimental trials
        print("\nBreath phase distribution (excluding practice):")
        if 'breathphase' in exp_df.columns:
            phase_counts = exp_df['breathphase'].value_counts().to_dict()
            for phase, count in sorted(phase_counts.items()):
                percent = (count / experimental_count) * 100
                print(f"  {phase}: {count} trials ({percent:.1f}%)")
        
        # 6. Cross-tabulation of trial types by breath phase
        print("\nTrial types by breath phase (COUNTERBALANCE CHECK):")
        if 'breathphase' in exp_df.columns:
            trial_by_phase = pd.crosstab(
                exp_df['trial_type'], 
                exp_df['breathphase']
            )
            print(trial_by_phase)
            
            # Calculate chi-square statistic for independence (excluding mandatory assignments)
            try:
                # Create subset without inhalation/exhalation (which are deterministically assigned)
                flex_df = exp_df[exp_df['trial_type'].isin(['baseline', 'catch'])]
                if len(flex_df) > 0:
                    flex_table = pd.crosstab(flex_df['trial_type'], flex_df['breathphase'])
                    chi2 = 0
                    row_sums = flex_table.sum(axis=1)
                    col_sums = flex_table.sum(axis=0) 
                    total = flex_table.sum().sum()
                    
                    for i in flex_table.index:
                        for j in flex_table.columns:
                            observed = flex_table.loc[i, j]
                            expected = row_sums[i] * col_sums[j] / total
                            if expected > 0:
                                chi2 += ((observed - expected) ** 2) / expected
                    
                    print(f"\nChi-square for baseline/catch distribution: {chi2:.2f}")
                    print(f"Lower values indicate better balance between breathphases.")
            except Exception as e:
                print(f"Error calculating chi-square: {e}")
        
        # 7. Consecutive trials analysis (excluding practice)
        trial_sequence = exp_df['trial_type'].tolist()
        consecutive_counts = {}
        
        if trial_sequence:
            current_type = trial_sequence[0]
            current_count = 1
            for t in trial_sequence[1:]:
                if t == current_type:
                    current_count += 1
                else:
                    key = f"{current_type}_{current_count}"
                    consecutive_counts[key] = consecutive_counts.get(key, 0) + 1
                    current_type = t
                    current_count = 1
            key = f"{current_type}_{current_count}"
            consecutive_counts[key] = consecutive_counts.get(key, 0) + 1
            
            print("\nRuns of consecutive trial types (excluding practice):")
            for key, count in sorted(consecutive_counts.items()):
                print(f"  {key}: {count}")
        
        # 8. SOA and stimulus type correlation check (excluding practice)
        print("\nSOA-by-stimulus-type COUNTERBALANCE CHECK (excluding practice):")
        non_baseline = exp_df[exp_df['trial_type'] != 'baseline']
        if len(non_baseline) > 0:
            soa_by_stim = pd.crosstab(
                non_baseline['soa_value_ms'], 
                non_baseline['stimulus_type']
            )
            print(soa_by_stim)
            
            # Calculate coefficient of variation as a balance metric
            cv_values = []
            for _, row in soa_by_stim.iterrows():
                if row.sum() > 0:
                    cv = row.std() / row.mean() if row.mean() > 0 else 0
                    cv_values.append(cv)
            
            mean_cv = sum(cv_values) / len(cv_values) if cv_values else 0
            print(f"\nCoefficient of variation across SOA-stimulus combinations: {mean_cv:.4f}")
            print(f"Lower values indicate better counterbalancing.")
        
        # 9. Breathphase alternation check
        if 'breathphase' in design_df.columns:
            phases = design_df.sort_values('milliseconds')['breathphase'].str.lower().tolist()
            alternation_violations = 0
            for i in range(1, len(phases)):
                if phases[i] == phases[i-1]:
                    alternation_violations += 1
            
            print(f"\nBreathphase alternation violations: {alternation_violations}")
            print(f"This should be 0 - the inhale/exhale cycle must be maintained.")
        
        # 10. Timing analysis
        if 'jittered_ms' in design_df.columns:
            time_diffs = []
            sorted_df = design_df.sort_values('jittered_ms')
            for i in range(1, len(sorted_df)):
                diff = sorted_df.iloc[i]['jittered_ms'] - sorted_df.iloc[i-1]['jittered_ms']
                time_diffs.append(diff)
            
            if time_diffs:
                min_diff = min(time_diffs)
                max_diff = max(time_diffs)
                avg_diff = sum(time_diffs) / len(time_diffs)
                print(f"\nStimulus timing intervals:")
                print(f"  Minimum interval: {min_diff:.1f} ms")
                print(f"  Maximum interval: {max_diff:.1f} ms")
                print(f"  Average interval: {avg_diff:.1f} ms")
        
        # 11. Practice trials validation
        if not practice_df.empty:
            print("\nPractice trials analysis:")
            practice_phase_counts = practice_df['breathphase'].value_counts().to_dict() if 'breathphase' in practice_df.columns else {}
            print(f"  Total practice trials: {len(practice_df)}")
            for phase, count in sorted(practice_phase_counts.items()):
                print(f"  {phase}: {count} trials")
            
            print(f"  Practice trial numbers: {practice_df['trial_number'].tolist()}")
            print(f"  Practice trials have jitter: {'jitter_ms' in practice_df.columns and not practice_df['jitter_ms'].isnull().all()}")
    
    # ------------------------------------------------------------------
    # Process a single participant: generate design, assign slots, finalize, and save CSV.
    # ------------------------------------------------------------------
    def process_participant(self, participant_id):
        try:
            if self.config['debug_mode']:
                print(f"\n=== Processing participant {participant_id} ===")
            design_df = self.generate_counterbalanced_design(participant_id)
            design_df = self.assign_breath_holds(design_df, participant_id)
            design_df = self.finalize_design_csv(design_df)
            out_path = os.path.join(self.experiment_log_dir, f"participant_{participant_id}_design.csv")
            design_df.to_csv(out_path, index=False)
            if self.config['debug_mode']:
                print(f"Saved participant {participant_id} design to {out_path}")
            return True, out_path
        except Exception as e:
            print(f"Error processing participant {participant_id}: {e}")
            traceback.print_exc()
            return False, None
    
    # ------------------------------------------------------------------
    # Run design generation for multiple participants.
    # ------------------------------------------------------------------
    def run(self, participant_ids=None, num_participants=None):
        if participant_ids is None and num_participants is None:
            num_participants = self.config.get('num_participants', 5)
        if participant_ids is None:
            participant_ids = list(range(num_participants))
        if self.config['debug_mode']:
            print("=== Starting PPSDesignGenerator with WAV Marker Integration ===")
            print(f"Parent WAV file: {self.marker_wav_file}")
            print(f"Parent CSV file: {self.marker_csv_file}")
            print(f"Marker count: {len(self.wav_markers)}")
            print(f"Repetitions: {self.config['repetitions']}")
            print(f"Practice trials: {self.config['practice_trials']}")
            print(f"Participants: {participant_ids}")
        results = {}
        for pid in participant_ids:
            success, path = self.process_participant(pid)
            results[pid] = {'success': success, 'csv_path': path}
        
        # Summary of all participants
        if self.config['debug_mode'] and len(results) > 1:
            print("\n=== CROSS-PARTICIPANT VALIDATION ===")
            all_success = all(info['success'] for info in results.values())
            print(f"All participants generated successfully: {all_success}")
        
        return results

# ----------------------------------------------------------------------
# MAIN FUNCTION
# ----------------------------------------------------------------------
def main():
    """Main function to run the design generator."""
    # Print configuration summary
    print("\n===== PPS EXPERIMENT DESIGN GENERATOR WITH WAV MARKER INTEGRATION =====")
    audio_dir = CONFIG['paths']['experiment_audio_dir']
    parent_wav = os.path.join(audio_dir, CONFIG['paths']['parent_wav_name'])
    parent_csv = os.path.join(audio_dir, CONFIG['paths']['parent_wav_name'].replace('.wav', '_markers.csv'))
    
    print(f"Audio directory: {audio_dir}")
    print(f"Parent WAV file: {parent_wav}")
    print(f"Parent CSV file: {parent_csv}")
    print(f"Output directory: {CONFIG['paths']['experiment_log_dir']}")
    print(f"Repetitions (trials per SOA per condition): {CONFIG['repetitions']}")
    print(f"Practice trials at beginning: {CONFIG['practice_trials']}")
    print(f"SOA Conditions: {CONFIG['soa_conditions_ms']}")
    print(f"Jitter Options: {CONFIG['jitter_options_ms']}")
    print(f"Looming Stimuli: {LOOMING_STIMULI}")
    print(f"Number of Participants: {CONFIG['num_participants']}")
    
    # Calculate expected trial counts
    reps = CONFIG['repetitions']
    soa_count = len(CONFIG['soa_conditions_ms'])
    
    trials_per_condition = reps * soa_count
    enabled_conditions = [cond for cond, enabled in CONFIG['included_conditions'].items() if enabled]
    total_base_trials = trials_per_condition * len(enabled_conditions)
    catch_trials = int(np.ceil(total_base_trials * CONFIG['catch_trial_percentage'] / 100))
    practice_trials = CONFIG['practice_trials']
    
    print("\n===== EXPECTED TRIAL COUNTS =====")
    print(f"Trials per condition: {reps} trials per SOA × {soa_count} SOAs = {trials_per_condition}")
    for cond in enabled_conditions:
        print(f"  {cond.upper()}: {trials_per_condition} trials")
    print(f"Catch trials ({CONFIG['catch_trial_percentage']}%): {catch_trials}")
    print(f"Practice trials: {practice_trials}")
    print(f"TOTAL TRIALS PER PARTICIPANT: {total_base_trials + catch_trials + practice_trials}")
    
    # Run the generator
    generator = PPSDesignGenerator(CONFIG)
    results = generator.run()
    
    # Summarize results
    print("\n===== GENERATION SUMMARY =====")
    success_count = sum(1 for info in results.values() if info['success'])
    print(f"Successfully generated designs for {success_count}/{len(results)} participants.")
    
    for pid, info in results.items():
        if info['success']:
            print(f"Participant {pid}: CSV generated at {info['csv_path']}")
        else:
            print(f"Participant {pid}: ERROR - no CSV generated.")

if __name__ == "__main__":
    main()
```

```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-
"""
Integrated PPS Design Generator

This script handles the entire process of:
1. Calculating required trial counts based on configuration
2. Creating the appropriate length WAV file with embedded breath markers
3. Extracting those markers to a CSV file
4. Using the markers as a scaffold to generate participant-specific designs

The workflow integrates the WAV marker embedding functionality with the
experimental design generation in a single coherent process.

Author: AI Assistant
"""

import os
import numpy as np
import pandas as pd
import random
import traceback
import wave
import struct
from pathlib import Path
import io
from pydub import AudioSegment

# ----------------------------------------------------------------------
# CONFIGURATION PARAMETERS
# ----------------------------------------------------------------------
CONFIG = {
    'repetitions': 12,           # Number of trials per SOA condition
    'num_participants': 100,     # Default number of participants
    'practice_trials': 8,        # Number of practice trials at the beginning
    'included_conditions': {
        'inhalation': True,
        'exhalation': True,
        'baseline': True,
    },
    'catch_trial_percentage': 10,
    'soa_conditions_ms': [190, 400, 700, 1000, 1500],
    'jitter_options_ms': [100, 200, 300, 400, 500],
    'debug_mode': True,
    'paths': {
        # Original Box Breathing audio files
        'intro_audio': r"C:\Users\cogpsy-vrlab\Documents\GitHub\BreathingSpace\Level0_StimulusFiles\BoxBreathing_intro.wav",
        'repetition_audio': r"C:\Users\cogpsy-vrlab\Documents\GitHub\BreathingSpace\Level0_StimulusFiles\BoxBreathing_repetitionsegment.wav",
        # Original timestamp CSVs
        'intro_csv': r"C:\Users\cogpsy-vrlab\Documents\GitHub\BreathingSpace\Level0_StimulusFiles\BoxBreathing_timestamps.csv",
        'repetition_csv': r"C:\Users\cogpsy-vrlab\Documents\GitHub\BreathingSpace\Level0_StimulusFiles\BoxBreathing_repetition_timestamps.csv",
        # Output directories
        'experiment_audio_dir': r"C:\Users\cogpsy-vrlab\Documents\GitHub\BreathingSpace\Level1_AudioGeneration\ExperimentAudio",
        'experiment_log_dir': r"C:\Users\cogpsy-vrlab\Documents\GitHub\BreathingSpace\Level1_AudioGeneration\ExperimentLog",
    },
    'audio': {
        'repetition_count': None,  # Will be calculated based on trial count
        'sample_rate': 48000,      # Audio sample rate in Hz
    }
}

# Noise types for looming stimuli (can be any labels)
LOOMING_STIMULI = ['blue', 'brown', 'pink', 'white']

# Marker type constants for WAV file marker embedding
MARKER_TYPE_INHALATION = 1  # Odd IDs: 1, 3, 5, 7, ...
MARKER_TYPE_EXHALATION = 2  # Even IDs: 2, 4, 6, 8, ...

# ----------------------------------------------------------------------
# Helper Functions
# ----------------------------------------------------------------------
def convert_ms_to_timestamp(ms):
    """Convert milliseconds to timestamp in MM:SS.s format."""
    if ms is None:
        return None
        
    m = int(ms // 60000)
    s = (ms % 60000) / 1000.0
    return f"{m:02}:{s:.1f}"

def read_markers_from_csv(csv_file):
    """Read marker timestamps and breathing phase info from a CSV file"""
    df = pd.read_csv(csv_file)
    print(f"CSV columns: {df.columns.tolist()}")
    
    markers = []
    
    # Check for breathphase column and expected format
    if 'breathphase' in df.columns:
        for _, row in df.iterrows():
            # Determine marker type based on breathphase
            marker_type = None
            if 'inhale' in str(row['breathphase']).lower():
                marker_type = MARKER_TYPE_INHALATION
            elif 'exhale' in str(row['breathphase']).lower():
                marker_type = MARKER_TYPE_EXHALATION
            else:
                print(f"Warning: Unknown breathing phase: {row['breathphase']}")
                continue  # Skip this marker if we can't classify it
            
            # Get required timestamp information
            marker = {
                'timestamp': row['milliseconds'] / 1000.0,  # Convert to seconds for WAV cue points
                'breathphase': row['breathphase'],
                'marker_type': marker_type
            }
            
            # Copy all CSV columns for later reference
            for col in df.columns:
                if col not in marker:
                    marker[col] = row[col]
            
            # Print a few examples of the original retentionslot_timestamp format
            if len(markers) < 3:
                print(f"Original timestamp format: {row['retentionslot_timestamp']}")
                    
            markers.append(marker)
    else:
        print(f"Error: CSV file {csv_file} does not contain a 'breathphase' column")
        return [], df.columns.tolist()
    
    # Sort markers by timestamp
    markers.sort(key=lambda x: x['timestamp'])
    return markers, df.columns.tolist()

def add_markers_to_wav(wav_file, markers, output_file):
    """Add inhale/exhale markers directly to WAV file using odd/even IDs"""
    # Read audio file
    audio = AudioSegment.from_wav(wav_file)
    sample_rate = audio.frame_rate
    
    # Create cue points list for the WAV file - ONLY encoding the breathphase
    cue_points = []
    
    # Write debug info
    print(f"Adding markers to {output_file}:")
    print(f"  Sample rate: {sample_rate} Hz")
    
    # Current ID counters for each type
    inhale_id = 1  # Starting with 1 (odd)
    exhale_id = 2  # Starting with 2 (even)
    
    for i, marker in enumerate(markers):
        # Convert timestamp to sample position
        sample_pos = int(marker['timestamp'] * sample_rate)
        
        # Get marker type (inhale or exhale) and assign ID
        marker_type = marker.get('marker_type')
        if marker_type == MARKER_TYPE_INHALATION:
            cue_id = inhale_id
            inhale_id += 2  # Next odd number
        else:  # MARKER_TYPE_EXHALATION
            cue_id = exhale_id
            exhale_id += 2  # Next even number
        
        # Debug output
        if i < 5 or i > len(markers) - 5 or i % 100 == 0:  # Show first/last 5 markers and every 100th
            print(f"  Marker {i+1}: ID={cue_id}, Position={sample_pos}, Type={'Inhale' if marker_type == MARKER_TYPE_INHALATION else 'Exhale'}")
        
        # Add to cue points - just the binary classification info
        cue_points.append((cue_id, sample_pos, 0, 0, 0, 0, 0, 0))
    
    # Save to WAV with embedded cue points
    _export_wav_with_markers(audio, output_file, cue_points)
    
    print(f"Created {output_file} with {len(markers)} embedded markers")
    return output_file

def _export_wav_with_markers(audio, output_file, cue_points):
    """Export a WAV file with embedded cue points"""
    # First export the audio to a temporary file
    temp_file = io.BytesIO()
    audio.export(temp_file, format="wav")
    temp_file.seek(0)
    
    # Read the WAV data
    with wave.open(temp_file, 'rb') as wf:
        params = wf.getparams()
        frames = wf.readframes(wf.getnframes())
    
    # Write the WAV file with cue points
    with wave.open(output_file, 'wb') as wf:
        wf.setparams(params)
        wf.writeframes(frames)
    
    # Add cue chunk if we have cue points
    if cue_points:
        try:
            # We need to manually add the cue chunk
            with open(output_file, 'rb+') as f:
                # Go to the end of the file
                f.seek(0, 2)
                file_size = f.tell()
                
                # Create cue chunk
                cue_data = struct.pack('<L', len(cue_points))  # Number of cue points
                
                for cue_id, position, chunk_start, block_start, sample_offset, _, _, _ in cue_points:
                    # Pack the cue point data
                    cue_data += struct.pack('<L', cue_id)  # cue ID
                    cue_data += struct.pack('<L', position)  # play position
                    cue_data += b'data'  # data chunk ID
                    cue_data += struct.pack('<L', 0)  # chunk start
                    cue_data += struct.pack('<L', 0)  # block start
                    cue_data += struct.pack('<L', 0)  # sample offset
                
                # Write cue chunk header
                f.write(b'cue ')
                f.write(struct.pack('<L', len(cue_data)))  # chunk size
                f.write(cue_data)  # chunk data
                
                # Update RIFF chunk size
                end_pos = f.tell()
                f.seek(4)
                f.write(struct.pack('<L', end_pos - 8))  # RIFF size = file size - 8
                
                print(f"Added cue chunk with {len(cue_points)} markers, total file size: {end_pos} bytes")
        except Exception as e:
            print(f"Error adding cue chunk: {e}")
            import traceback
            traceback.print_exc()
    
    # Verify the file has the expected markers
    try:
        with open(output_file, 'rb') as f:
            # Check for cue chunk
            f.seek(12)  # Skip RIFF header and file size
            
            while True:
                chunk_pos = f.tell()
                chunk_id = f.read(4)
                
                if not chunk_id or len(chunk_id) < 4:
                    break
                
                chunk_size = int.from_bytes(f.read(4), byteorder='little')
                
                if chunk_id == b'cue ':
                    num_cues = int.from_bytes(f.read(4), byteorder='little')
                    print(f"Verified: WAV file has {num_cues} cue points")
                    break
                
                # Skip to next chunk
                f.seek(chunk_pos + 8 + chunk_size)
    except Exception as e:
        print(f"Error verifying cue chunk: {e}")
    
    return output_file

def extract_markers_from_wav(wav_file):
    """
    Extract marker information directly from a WAV file's cue points 
    using odd/even ID detection.
    """
    # Read the audio file to get the sample rate
    audio = AudioSegment.from_wav(wav_file)
    sample_rate = audio.frame_rate
    
    markers = []
    
    try:
        with open(wav_file, 'rb') as f:
            # Check RIFF header
            if f.read(4) != b'RIFF':
                raise ValueError("Not a valid RIFF file")
            
            # Read file size (and skip)
            f.read(4)
            
            # Check WAVE format
            if f.read(4) != b'WAVE':
                raise ValueError("Not a valid WAVE file")
            
            # Find all chunks, including potentially multiple cue chunks
            while True:
                chunk_pos = f.tell()
                chunk_header = f.read(4)
                
                if not chunk_header or len(chunk_header) < 4:
                    break  # End of file
                
                chunk_size = int.from_bytes(f.read(4), byteorder='little')
                print(f"Found chunk: {chunk_header.decode('ascii', errors='replace')} size: {chunk_size}")
                
                if chunk_header == b'cue ':
                    # Found cue chunk
                    num_cue_points = int.from_bytes(f.read(4), byteorder='little')
                    print(f"Found {num_cue_points} cue points in WAV file")
                    
                    # Read all cue points
                    for i in range(num_cue_points):
                        cue_chunk_start = f.tell()
                        
                        # Read cue point
                        cue_id_bytes = f.read(4)
                        cue_id = int.from_bytes(cue_id_bytes, byteorder='little')
                        
                        position_bytes = f.read(4)
                        position = int.from_bytes(position_bytes, byteorder='little')
                        
                        data_chunk_id = f.read(4)  # Should be 'data'
                        chunk_start = int.from_bytes(f.read(4), byteorder='little')
                        block_start = int.from_bytes(f.read(4), byteorder='little')
                        sample_offset = int.from_bytes(f.read(4), byteorder='little')
                        
                        if i < 5 or i > num_cue_points - 5 or i % 100 == 0:
                            print(f"  Cue point {i+1}: ID={cue_id}, Position={position}, DataChunk={data_chunk_id}")
                        
                        # Determine marker type from cue ID parity (odd/even)
                        marker_type = None
                        breathphase = None
                        
                        if cue_id % 2 == 1:  # Odd ID = Inhale
                            marker_type = MARKER_TYPE_INHALATION
                            breathphase = "inhale"
                        elif cue_id % 2 == 0:  # Even ID = Exhale
                            marker_type = MARKER_TYPE_EXHALATION
                            breathphase = "exhale"
                        else:
                            # This condition can't actually happen (every number is either odd or even),
                            # but kept for consistency with previous logic
                            print(f"  Skipping cue ID {cue_id} - unexpected value")
                            continue
                        
                        # Calculate timestamp
                        timestamp_seconds = position / sample_rate
                        milliseconds = int(timestamp_seconds * 1000)
                        
                        # Format retentionslot_timestamp in MM:SS.S format
                        minutes = int(timestamp_seconds // 60)
                        seconds = timestamp_seconds % 60
                        retentionslot_timestamp = f"{minutes:02}:{seconds:.1f}"
                        
                        # Create marker entry with timestamps in required formats
                        marker = {
                            'retentionslot_timestamp': retentionslot_timestamp,
                            'milliseconds': milliseconds,
                            'sample_index': position,
                            'description': "",  # Empty as requested
                            'breathphase': breathphase,
                            'marker_type': marker_type
                        }
                        
                        markers.append(marker)
                    
                    # Continue searching for more cue chunks (don't break)
                    f.seek(chunk_pos + 8 + chunk_size)
                else:
                    # Skip this chunk
                    f.seek(chunk_size, 1)  # 1 means relative to current position
        
    except Exception as e:
        print(f"Error reading WAV file: {e}")
        import traceback
        traceback.print_exc()
    
    # Sort markers by timestamp
    markers.sort(key=lambda x: x['milliseconds'])
    
    if not markers:
        print("No markers found in WAV file")
    else:
        print(f"Extracted {len(markers)} markers from WAV file")
        print("Sample markers:")
        for i in range(min(5, len(markers))):
            print(f"  {i+1}: time={markers[i]['retentionslot_timestamp']}, phase={markers[i]['breathphase']}")
    
    return markers

def export_markers_to_csv(markers, output_csv, column_format=None):
    """Export markers to a CSV file in the specified format"""
    # Create a DataFrame with the specified columns or all available columns
    df_data = []
    
    for marker in markers:
        row = {}
        if column_format:
            # Use only specified columns
            for col in column_format:
                row[col] = marker.get(col, "")
        else:
            # Use all available columns
            row = marker.copy()
        df_data.append(row)
    
    # Create DataFrame
    if column_format:
        df = pd.DataFrame(df_data, columns=column_format)
    else:
        df = pd.DataFrame(df_data)
    
    # Save to CSV
    df.to_csv(output_csv, index=False)
    print(f"Exported {len(markers)} markers to {output_csv}")
    return output_csv

def adjust_markers_for_concatenated_file(markers1, markers2, duration1, sample_rate, repetition_count=1):
    """
    Adjust markers from the second file for multiple repetitions and
    combine with markers from the first file.
    """
    # Start with markers from the first file
    all_markers = markers1.copy()
    
    # Duration of the second segment
    duration2 = markers2[-1]['timestamp'] - markers2[0]['timestamp'] + 1.0  # Add buffer
    
    # Add markers from each repeat of the second file
    current_offset = duration1
    for i in range(repetition_count):
        # Adjust markers from this repeat
        for marker in markers2:
            adjusted_marker = marker.copy()
            
            # Calculate new timestamp in seconds
            adjusted_timestamp = marker['timestamp'] + current_offset
            
            # Update all timestamp fields
            adjusted_marker['timestamp'] = adjusted_timestamp
            adjusted_marker['milliseconds'] = int(adjusted_timestamp * 1000)
            adjusted_marker['sample_index'] = int(adjusted_timestamp * sample_rate)
            
            # Format retention timestamp
            adjusted_marker['retentionslot_timestamp'] = convert_ms_to_timestamp(adjusted_marker['milliseconds'])
            
            all_markers.append(adjusted_marker)
        
        # Update the offset for the next repeat
        current_offset += duration2
    
    # Sort all markers by timestamp
    all_markers.sort(key=lambda x: x['timestamp'])
    return all_markers

def create_concatenated_audio_with_markers(config):
    """
    Create a concatenated audio file with embedded markers based on the
    required trial count.
    
    Args:
        config: Configuration dictionary with paths and parameters
        
    Returns:
        Tuple of (concatenated_wav_path, marker_csv_path, all_markers)
    """
    # Prepare paths
    intro_wav = config['paths']['intro_audio']
    repetition_wav = config['paths']['repetition_audio']
    intro_csv = config['paths']['intro_csv']
    repetition_csv = config['paths']['repetition_csv']
    output_dir = config['paths']['experiment_audio_dir']
    
    # Create output directory if it doesn't exist
    os.makedirs(output_dir, exist_ok=True)
    
    # Determine repetition count
    repetition_count = config['audio']['repetition_count']
    repeat_suffix = f"_x{repetition_count}" if repetition_count > 1 else ""
    
    # Output file paths
    wav_with_markers1 = os.path.join(output_dir, "BoxBreathing_intro_with_markers.wav")
    wav_with_markers2 = os.path.join(output_dir, "BoxBreathing_repetitionsegment_with_markers.wav")
    concatenated_wav = os.path.join(output_dir, f"BoxBreathing_complete{repeat_suffix}.wav")
    output_csv = os.path.join(output_dir, f"BoxBreathing_complete{repeat_suffix}_markers.csv")
    
    print(f"Configuration: Repeating second segment {repetition_count} times")
    
    # STEP 1: Read markers from CSV files
    print(f"Reading markers from {intro_csv}...")
    markers1, columns1 = read_markers_from_csv(intro_csv)
    print(f"Found {len(markers1)} markers in first CSV")
    
    print(f"Reading markers from {repetition_csv}...")
    markers2, columns2 = read_markers_from_csv(repetition_csv)
    print(f"Found {len(markers2)} markers in second CSV")
    
    # Get the CSV column format
    csv_columns = columns1 if 'breathphase' in columns1 else columns2
    
    # STEP 2: Create new WAV files with embedded markers
    print(f"Creating new WAV file with markers: {wav_with_markers1}")
    add_markers_to_wav(intro_wav, markers1, wav_with_markers1)
    
    print(f"Creating new WAV file with markers: {wav_with_markers2}")
    add_markers_to_wav(repetition_wav, markers2, wav_with_markers2)
    
    # STEP 3: Concatenate the WAV files with repetition
    print(f"Concatenating WAV files to: {concatenated_wav} (with {repetition_count} repeats)")
    
    # Read the audio segments
    audio1 = AudioSegment.from_wav(wav_with_markers1)
    audio2 = AudioSegment.from_wav(wav_with_markers2)
    
    # Get duration of first segment
    duration1 = len(audio1) / 1000.0  # pydub uses milliseconds
    sample_rate = audio1.frame_rate
    
    # Start with the first segment
    concatenated = audio1
    
    # Add repeated instances of the second segment
    for i in range(repetition_count):
        print(f"  Adding repeat {i+1} of {repetition_count}...")
        concatenated += audio2
    
    # Export the concatenated audio
    concatenated.export(concatenated_wav, format="wav")
    print(f"Created concatenated file: {concatenated_wav}")
    
    # STEP 4: Combine and adjust all markers
    print(f"Adjusting markers for concatenated file with {repetition_count} repeats...")
    all_markers = adjust_markers_for_concatenated_file(
        markers1, markers2, duration1, sample_rate, repetition_count)
    
    # STEP 5: Add all markers to the concatenated file
    print(f"Adding all markers to concatenated file...")
    add_markers_to_wav(concatenated_wav, all_markers, concatenated_wav)
    
    # STEP 6: Extract markers directly from the concatenated WAV file
    print(f"Extracting markers from concatenated WAV file...")
    extracted_markers = extract_markers_from_wav(concatenated_wav)
    
    # STEP 7: Export the extracted markers to a CSV in the original format
    print(f"Exporting extracted markers to CSV: {output_csv}")
    export_markers_to_csv(extracted_markers, output_csv, csv_columns)
    
    # Print summary
    print("\nAudio Creation Complete!")
    print(f"Created WAV files with embedded markers:")
    print(f"  - {wav_with_markers1}")
    print(f"  - {wav_with_markers2}")
    print(f"Concatenated file with all markers ({repetition_count} repeats):")
    print(f"  - {concatenated_wav}")
    print(f"Extracted markers to CSV:")
    print(f"  - {output_csv}")
    
    # Print marker summary
    inhalation_count = sum(1 for m in extracted_markers if m.get('marker_type') == MARKER_TYPE_INHALATION)
    exhalation_count = sum(1 for m in extracted_markers if m.get('marker_type') == MARKER_TYPE_EXHALATION)
    
    print("\nExtracted breathing phase marker summary:")
    print(f"  Inhale markers: {inhalation_count}")
    print(f"  Exhale markers: {exhalation_count}")
    print(f"  Total markers: {len(extracted_markers)}")
    print(f"  Expected markers: {len(markers1) + (len(markers2) * repetition_count)}")
    
    return concatenated_wav, output_csv, extracted_markers

# ----------------------------------------------------------------------
# CLASS: PPSDesignGenerator
# ----------------------------------------------------------------------
class PPSDesignGenerator:
    """Generates participant-specific CSV design files for PPS experiments."""
    
    def __init__(self, config=None, wav_markers=None):
        self.config = config or CONFIG
        
        # Prepare output directory
        self.experiment_log_dir = self.config['paths']['experiment_log_dir']
        os.makedirs(self.experiment_log_dir, exist_ok=True)
        
        # Calculate trial counts
        self.trial_counts = self._calculate_trial_counts()
        self.conditions = [cond for cond, inc in self.config['included_conditions'].items() if inc]
        
        # Set seed for reproducibility
        self.base_seed = 42
        
        # Use provided markers or create artificial scaffold
        self.wav_markers = wav_markers or []
        
        # Default column structure that must be preserved
        self.scaffold_columns = [
            'retentionslot_timestamp', 'milliseconds', 'sample_index', 
            'description', 'breathphase'
        ]
        
        if self.config['debug_mode']:
            total_trials = sum(self.trial_counts.values())
            print("Initialized PPSDesignGenerator with:")
            print(f"  - Repetitions: {self.config['repetitions']}")
            print(f"  - Practice trials: {self.config['practice_trials']}")
            print(f"  - Conditions: {self.conditions}")
            print(f"  - SOA values: {self.config['soa_conditions_ms']}")
            print(f"  - Total trials per participant: {total_trials} + {self.config['practice_trials']} practice = {total_trials + self.config['practice_trials']}")
            print(f"  - Output directory: {self.experiment_log_dir}")
            print(f"  - WAV markers: {len(self.wav_markers)}")
    
    def _calculate_trial_counts(self):
        """
        Calculate the number of trials for each condition, using the repetitions parameter
        to determine how many trials per SOA per condition.
        """
        soa_conditions = self.config['soa_conditions_ms']
        repetitions = self.config['repetitions']
        catch_percentage = self.config['catch_trial_percentage']
        
        trial_counts = {}
        for condition, include in self.config['included_conditions'].items():
            if include:
                # Each condition gets repetitions × number of SOAs trials
                trial_counts[condition] = len(soa_conditions) * repetitions
        
        regular_trials = sum(trial_counts.values())
        catch_trials = int(np.ceil(regular_trials * catch_percentage / 100))
        trial_counts['catch'] = catch_trials
        
        if self.config['debug_mode']:
            print("Trial count calculation:")
            for cond, count in trial_counts.items():
                print(f"  {cond}: {count} trials")
            print(f"  Total: {sum(trial_counts.values())} trials")
        
        return trial_counts

    def calculate_required_repetitions(self):
        """
        Calculate the number of repetition segments needed for the audio file
        based on trial counts and practice trials.
        
        Returns:
            Integer number of required repetitions
        """
        # Calculate total trial count
        total_trials = sum(self.trial_counts.values()) + self.config.get('practice_trials', 0)
        
        # Estimate slots per segment
        # Read marker counts from original CSVs
        intro_csv = self.config['paths']['intro_csv']
        repetition_csv = self.config['paths']['repetition_csv']
        
        try:
            intro_markers, _ = read_markers_from_csv(intro_csv)
            repetition_markers, _ = read_markers_from_csv(repetition_csv)
            
            intro_slots = len(intro_markers)
            repetition_slots = len(repetition_markers)
            
            print(f"Intro has {intro_slots} slots, repetition segment has {repetition_slots} slots")
            
            # Determine how many repetition segments are needed
            if total_trials <= intro_slots:
                # Intro segment has enough slots alone
                repetitions_needed = 0
            else:
                # Need repetition segments for remaining trials
                remaining_trials = total_trials - intro_slots
                repetitions_needed = (remaining_trials + repetition_slots - 1) // repetition_slots  # Ceiling division
            
            print(f"Need {repetitions_needed} repetition segments for {total_trials} total trials")
            return max(1, repetitions_needed)  # At least 1 repetition
            
        except Exception as e:
            print(f"Error calculating repetitions: {e}")
            # Default fallback
            return 5  # Default to 5 repetitions if can't calculate

    # ------------------------------------------------------------------
    # Create scaffold from WAV markers or fallback to artificial
    # ------------------------------------------------------------------
    def create_retention_slot_scaffold(self, required_slots):
        """
        Create a scaffold DataFrame from WAV markers or fallback to artificial.
        
        Args:
            required_slots: Minimum number of slots needed
            
        Returns:
            DataFrame with scaffold structure
        """
        if self.wav_markers and len(self.wav_markers) > 0:
            # Use extracted WAV markers as scaffold
            scaffold_df = pd.DataFrame(self.wav_markers)
            
            # Ensure all required columns exist
            for col in self.scaffold_columns:
                if col not in scaffold_df.columns:
                    scaffold_df[col] = None
                    print(f"Warning: Added missing column '{col}' to marker scaffold")
            
            print(f"Created scaffold from WAV markers: {len(scaffold_df)} slots")
            
        else:
            # Fallback: Create scaffold from scratch
            print("No WAV markers available. Creating artificial scaffold...")
            cycle_sec = 8.0  # Default breath cycle duration
            sample_rate = self.config.get('audio', {}).get('sample_rate', 48000)
            
            scaffold_data = []
            current_ms = 0
            
            for i in range(required_slots):
                # Alternate breathphase
                breathphase = "inhale" if i % 2 == 0 else "exhale"
                
                # Convert to timestamp format
                timestamp = convert_ms_to_timestamp(current_ms)
                
                # Create marker entry
                marker = {
                    'retentionslot_timestamp': timestamp,
                    'milliseconds': current_ms,
                    'sample_index': int(current_ms * sample_rate / 1000),
                    'description': f"Artificial {breathphase} marker {i+1}",
                    'breathphase': breathphase,
                    'marker_type': MARKER_TYPE_INHALATION if breathphase == "inhale" else MARKER_TYPE_EXHALATION
                }
                
                scaffold_data.append(marker)
                current_ms += cycle_sec * 1000  # Add 8 seconds for next marker
            
            scaffold_df = pd.DataFrame(scaffold_data)
            print(f"Created artificial scaffold with {len(scaffold_df)} slots")
        
        # Ensure we have enough slots
        if len(scaffold_df) < required_slots:
            # Need to extend scaffold
            print(f"Extending scaffold from {len(scaffold_df)} to {required_slots} slots")
            
            last_row = scaffold_df.iloc[-1]
            last_ms = float(last_row['milliseconds'])
            last_sample_index = int(last_row['sample_index'])
            cycle_sec = 8.0  # Default breath cycle duration
            sample_rate = self.config.get('audio', {}).get('sample_rate', 48000)
            
            # Get the last breathphase to ensure proper alternation
            last_breathphase = last_row['breathphase'].lower()
            
            # Create additional rows
            new_rows = []
            for i in range(required_slots - len(scaffold_df)):
                # Increment by cycle duration for each new row
                new_ms = last_ms + (cycle_sec * 1000) * (i + 1)
                new_sample_index = last_sample_index + int(cycle_sec * sample_rate * (i + 1))
                new_timestamp = convert_ms_to_timestamp(new_ms)
                
                # Alternate breathphase
                if i == 0:  # First new row
                    new_breathphase = 'inhale' if last_breathphase == 'exhale' else 'exhale'
                else:  # Subsequent rows
                    prev_phase = new_rows[i-1]['breathphase'].lower()
                    new_breathphase = 'inhale' if prev_phase == 'exhale' else 'exhale'
                
                # Create marker entry
                new_row = {
                    'retentionslot_timestamp': new_timestamp,
                    'milliseconds': new_ms,
                    'sample_index': new_sample_index,
                    'description': f"Extended {new_breathphase} marker",
                    'breathphase': new_breathphase,
                    'marker_type': MARKER_TYPE_INHALATION if new_breathphase == "inhale" else MARKER_TYPE_EXHALATION
                }
                
                new_rows.append(new_row)
            
            # Append new rows to scaffold
            if new_rows:
                scaffold_df = pd.concat([scaffold_df, pd.DataFrame(new_rows)], ignore_index=True)
        
        # Trim if we have too many slots
        if len(scaffold_df) > required_slots:
            scaffold_df = scaffold_df.iloc[:required_slots]
        
        # Ensure scaffold is sorted by time
        scaffold_df = scaffold_df.sort_values(by='milliseconds').reset_index(drop=True)
        
        return scaffold_df

    # ------------------------------------------------------------------
    # Generate counterbalanced design (unchanged)
    # ------------------------------------------------------------------
    def generate_counterbalanced_design(self, participant_id):
        """
        Generate a counterbalanced design for a participant with trials per SOA
        determined by the repetitions parameter in the config.
        
        Catch trials don't have SOA values assigned since they don't have tactile stimuli.
        """
        random.seed(self.base_seed + participant_id)
        np.random.seed(self.base_seed + participant_id)
        
        soa_conditions = self.config['soa_conditions_ms']
        jitter_options = self.config['jitter_options_ms']
        stimulus_types = LOOMING_STIMULI
        repetitions = self.config['repetitions']  # Trials per SOA per condition
        
        all_trials = []
        
        # Generate main conditions: inhalation, exhalation, and baseline
        for condition, count in self.trial_counts.items():
            if condition == 'catch' or count == 0:
                continue
            
            # For each SOA, generate exactly 'repetitions' trials
            for soa in soa_conditions:
                # Calculate how many trials per stimulus type (as evenly as possible)
                trials_per_soa = repetitions
                trials_per_stim = trials_per_soa // len(stimulus_types)  # Integer division
                remainder = trials_per_soa % len(stimulus_types)
                
                # Distribute trials across stimulus types
                stim_trials = []
                for i, stim_type in enumerate(stimulus_types):
                    # Add extra trial for remainder distribution
                    extra = 1 if i < remainder else 0
                    for _ in range(trials_per_stim + extra):
                        stim_trials.append(stim_type)
                
                # Shuffle the stimulus types
                random.shuffle(stim_trials)
                
                # Create trials for this SOA
                for stim_type in stim_trials:
                    all_trials.append({
                        'participant_id': participant_id,
                        'trial_type': condition,
                        'stimulus_type': stim_type,
                        'soa_value_ms': soa,
                        'jitter_ms': random.choice(jitter_options),
                        'is_tactile': True  # For these main trials, tactile is present.
                    })
        
        # Add catch trials (looming only, no tactile) - no SOA values needed
        catch_count = self.trial_counts.get('catch', 0)
        if catch_count > 0:
            # Distribute catch trials evenly across stimulus types only
            catch_trials = []
            stim_counts = {stim: 0 for stim in stimulus_types}
            
            for _ in range(catch_count):
                # Choose the stimulus type with the lowest count
                min_stim = min(stim_counts, key=stim_counts.get)
                
                catch_trials.append({
                    'participant_id': participant_id,
                    'trial_type': 'catch',
                    'stimulus_type': min_stim,
                    'soa_value_ms': None,  # No SOA for catch trials
                    'jitter_ms': random.choice(jitter_options),
                    'is_tactile': False
                })
                
                stim_counts[min_stim] += 1
            
            # Shuffle catch trials and add to all trials
            random.shuffle(catch_trials)
            all_trials.extend(catch_trials)
        
        # Shuffle all trials and assign trial numbers
        all_trials_df = pd.DataFrame(all_trials)
        all_trials_df = all_trials_df.sample(frac=1, random_state=participant_id*100).reset_index(drop=True)
        all_trials_df['trial_number'] = range(1, len(all_trials_df) + 1)
        
        if self.config['debug_mode']:
            total_trials = len(all_trials_df)
            print("\nTrial distribution:")
            print(f"Total trials: {total_trials}")
            
            # Count by trial type
            for trial_type in sorted(all_trials_df['trial_type'].unique()):
                type_df = all_trials_df[all_trials_df['trial_type'] == trial_type]
                print(f"Total {trial_type}: {len(type_df)}")
                
                # Count by SOA for non-baseline and non-catch trial types
                if trial_type not in ['baseline', 'catch']:
                    for soa in sorted(soa_conditions):
                        soa_count = len(type_df[type_df['soa_value_ms'] == soa])
                        print(f"  {trial_type}_{soa}: {soa_count}")
                elif trial_type == 'baseline':
                    # For baseline, show SOA distribution (though not used for looming)
                    for soa in sorted(soa_conditions):
                        soa_count = len(type_df[type_df['soa_value_ms'] == soa])
                        print(f"  baseline_{soa}: {soa_count}")
                elif trial_type == 'catch':
                    # For catch trials, show stimulus distribution
                    catch_count = len(type_df)
                    print(f"  catch_trials: {catch_count} (no tactile stimulus, no SOA)")
            
            # Show stimulus type distribution
            print("\nStimulus type distribution by trial type:")
            for trial_type in sorted(all_trials_df['trial_type'].unique()):
                if trial_type == 'baseline':
                    continue
                subset = all_trials_df[all_trials_df['trial_type'] == trial_type]
                print(f"  {trial_type}: {subset['stimulus_type'].value_counts().to_dict()}")
        
        return all_trials_df

    # ------------------------------------------------------------------
    # Assign scaffold retention slots to trials 
    # ------------------------------------------------------------------
    def assign_breath_holds(self, design_df, participant_id):
        """
        Assign scaffold retention slots to trials while strictly maintaining the breathphase cycle.
        
        Key requirements:
        1. Breathphase cycle (inhale-exhale-inhale-exhale) must remain intact
        2. Inhalation trials must go to inhale breathphase slots
        3. Exhalation trials must go to exhale breathphase slots
        4. Baseline and catch trials must be equally distributed between inhale and exhale breathphases
        5. The first N slots (where N is practice_trials) are reserved for practice trials
        """
        # Account for practice trials in required slots calculation
        practice_trials = self.config.get('practice_trials', 0)
        required_slots = len(design_df) + practice_trials
        scaffold_df = self.create_retention_slot_scaffold(required_slots)
        
        # Reserve the first N slots for practice trials
        practice_scaffold = scaffold_df.iloc[:practice_trials].copy() if practice_trials > 0 else pd.DataFrame()
        regular_scaffold = scaffold_df.iloc[practice_trials:].copy()
        
        # Create practice trials (empty placeholder trials)
        practice_trial_list = []
        if practice_trials > 0:
            for i, (_, row) in enumerate(practice_scaffold.iterrows()):
                breathphase = row['breathphase'].lower()
                practice_trial = {
                    'participant_id': participant_id,
                    'trial_number': i + 1,  # Start numbering from 1
                    'trial_type': 'practice',
                    'stimulus_type': None,
                    'soa_value_ms': None,
                    'jitter_ms': random.choice(self.config['jitter_options_ms']),
                    'is_tactile': False
                }
                # Add all scaffold columns
                for col in row.index:
                    practice_trial[col] = row[col]
                
                practice_trial_list.append(practice_trial)
        
        # Partition design trials by trial type
        pool_inh = [d for d in design_df.to_dict('records') if d['trial_type'] == 'inhalation']
        pool_exh = [d for d in design_df.to_dict('records') if d['trial_type'] == 'exhalation']
        pool_baseline = [d for d in design_df.to_dict('records') if d['trial_type'] == 'baseline']
        pool_catch = [d for d in design_df.to_dict('records') if d['trial_type'] == 'catch']
        
        # Count scaffold slots by breathphase
        inhale_slots = regular_scaffold[regular_scaffold['breathphase'].str.lower() == 'inhale']
        exhale_slots = regular_scaffold[regular_scaffold['breathphase'].str.lower() == 'exhale']
        n_inhale_slots = len(inhale_slots)
        n_exhale_slots = len(exhale_slots)
        
        if self.config['debug_mode']:
            print(f"Scaffold: {n_inhale_slots} inhale slots, {n_exhale_slots} exhale slots")
            print(f"Trials: {len(pool_inh)} inhalation, {len(pool_exh)} exhalation, " 
                  f"{len(pool_baseline)} baseline, {len(pool_catch)} catch")
            print(f"Practice trials: {practice_trials}")
        
        # Verify we have enough slots for mandatory trials
        if len(pool_inh) > n_inhale_slots:
            raise ValueError(f"Not enough inhale slots ({n_inhale_slots}) for inhalation trials ({len(pool_inh)})")
        if len(pool_exh) > n_exhale_slots:
            raise ValueError(f"Not enough exhale slots ({n_exhale_slots}) for exhalation trials ({len(pool_exh)})")
        
        # Calculate remaining slots per breathphase
        remaining_inhale = n_inhale_slots - len(pool_inh)
        remaining_exhale = n_exhale_slots - len(pool_exh)
        
        # Total baseline+catch trials
        total_flex_trials = len(pool_baseline) + len(pool_catch)
        
        # Verify we have enough slots for all trials
        if remaining_inhale + remaining_exhale != total_flex_trials:
            raise ValueError(f"Mismatch between available slots ({remaining_inhale + remaining_exhale}) "
                            f"and flexible trials ({total_flex_trials})")
        
        # Ensure equal distribution of baseline and catch trials between breathphases
        
        # Shuffle baseline and catch trials
        random.shuffle(pool_baseline)
        random.shuffle(pool_catch)
        
        # Calculate how many of each should go to each breathphase
        baseline_inhale = len(pool_baseline) // 2
        baseline_exhale = len(pool_baseline) - baseline_inhale
        
        catch_inhale = len(pool_catch) // 2
        catch_exhale = len(pool_catch) - catch_inhale
        
        # Create inhale and exhale flexible trial pools
        inhale_flexible = pool_baseline[:baseline_inhale] + pool_catch[:catch_inhale]
        exhale_flexible = pool_baseline[baseline_inhale:] + pool_catch[catch_inhale:]
        
        random.shuffle(inhale_flexible)
        random.shuffle(exhale_flexible)
        
        if self.config['debug_mode']:
            print(f"Distributing flexible trials:")
            print(f"  Baseline: {baseline_inhale} to inhale, {baseline_exhale} to exhale")
            print(f"  Catch: {catch_inhale} to inhale, {catch_exhale} to exhale")
        
        # Create final allocation pools
        inhale_pool = pool_inh + inhale_flexible
        exhale_pool = pool_exh + exhale_flexible
        
        # Shuffle within each pool to randomize order
        random.shuffle(inhale_pool)
        random.shuffle(exhale_pool)
        
        # Create a dictionary mapping breathphase to trial pool
        phase_to_pool = {
            'inhale': inhale_pool,
            'exhale': exhale_pool
        }
        
        # Initialize counters for each breathphase
        pool_counters = {
            'inhale': 0,
            'exhale': 0
        }
        
        # Final list to store assigned trials in scaffold order
        regular_assignments = []
        
        # Assign trials to regular scaffold rows, maintaining the original order
        for _, row in regular_scaffold.iterrows():
            breathphase = row['breathphase'].lower()
            
            # Get the next trial from the appropriate pool
            if pool_counters[breathphase] < len(phase_to_pool[breathphase]):
                trial = phase_to_pool[breathphase][pool_counters[breathphase]]
                pool_counters[breathphase] += 1
                
                # Combine scaffold information with trial information
                combined = trial.copy()
                for col in row.index:
                    combined[col] = row[col]
                
                regular_assignments.append(combined)
        
        # Combine practice trials and regular trials
        final_assignments = practice_trial_list + regular_assignments
        
        # Create final dataframe and sort by scaffold order
        final_df = pd.DataFrame(final_assignments).sort_values(by='milliseconds').reset_index(drop=True)
        
        # Renumber trials from 1 to N to ensure sequential numbering
        final_df['trial_number'] = range(1, len(final_df) + 1)
        
        # Compute additional timing information
        final_df['jittered_ms'] = final_df['milliseconds'] + final_df['jitter_ms'].fillna(0)
        
        # Add looming stimulus timestamp for non-baseline trials
        looming_mask = ~final_df['trial_type'].isin(['baseline', 'practice'])
        final_df.loc[looming_mask, 'looming_stimulus_timestamp'] = final_df.loc[looming_mask, 'jittered_ms'].apply(convert_ms_to_timestamp)
        
        # Add SOA timing for non-catch, non-practice trials
        non_catch_mask = ~final_df['trial_type'].isin(['catch', 'practice'])
        final_df.loc[non_catch_mask, 'soa_ms'] = final_df.loc[non_catch_mask, 'jittered_ms'] + final_df.loc[non_catch_mask, 'soa_value_ms'].fillna(0)
        final_df.loc[non_catch_mask, 'tactile_stimulus_timestamp'] = final_df.loc[non_catch_mask, 'soa_ms'].apply(convert_ms_to_timestamp)
        
        # Verify the assignment respects the breathphase constraints
        self._validate_breathphase_assignment(final_df)
        
        return final_df

    def _validate_breathphase_assignment(self, final_df):
        """Verify that the trial assignment respects breathphase constraints."""
        if not self.config['debug_mode']:
            return
            
        print("\n=== BREATHPHASE ASSIGNMENT VALIDATION ===")
        
        # Skip practice trials for validation
        non_practice_df = final_df[final_df['trial_type'] != 'practice']
        
        # Check for inhalation trials assigned to inhale
        inh_trials = non_practice_df[non_practice_df['trial_type'] == 'inhalation']
        inh_correct = (inh_trials['breathphase'].str.lower() == 'inhale').all() if len(inh_trials) > 0 else True
        print(f"Inhalation trials on inhale breathphase: {inh_correct}")
        
        # Check for exhalation trials assigned to exhale
        exh_trials = non_practice_df[non_practice_df['trial_type'] == 'exhalation']
        exh_correct = (exh_trials['breathphase'].str.lower() == 'exhale').all() if len(exh_trials) > 0 else True
        print(f"Exhalation trials on exhale breathphase: {exh_correct}")
        
        # Count distribution of baseline and catch trials across breathphases
        baseline_inhale = ((non_practice_df['trial_type'] == 'baseline') & 
                         (non_practice_df['breathphase'].str.lower() == 'inhale')).sum()
        baseline_exhale = ((non_practice_df['trial_type'] == 'baseline') & 
                          (non_practice_df['breathphase'].str.lower() == 'exhale')).sum()
        catch_inhale = ((non_practice_df['trial_type'] == 'catch') & 
                       (non_practice_df['breathphase'].str.lower() == 'inhale')).sum()
        catch_exhale = ((non_practice_df['trial_type'] == 'catch') & 
                       (non_practice_df['breathphase'].str.lower() == 'exhale')).sum()
        
        print(f"Baseline trials: {baseline_inhale} on inhale, {baseline_exhale} on exhale")
        print(f"Catch trials: {catch_inhale} on inhale, {catch_exhale} on exhale")
        
        # Verify breathphase cycle is maintained
        prev_phase = None
        cycle_maintained = True
        expected_cycle = ['inhale', 'exhale'] * (len(final_df) // 2 + 1)
        
        for i, phase in enumerate(final_df['breathphase'].str.lower()):
            if i < len(expected_cycle) and phase != expected_cycle[i]:
                cycle_maintained = False
                print(f"Breathphase cycle disrupted at index {i}: expected {expected_cycle[i]}, got {phase}")
                break
        
        print(f"Breathphase cycle maintained: {cycle_maintained}")
        
        # Count practice trials
        practice_count = (final_df['trial_type'] == 'practice').sum()
        print(f"Practice trials: {practice_count}")

    # ------------------------------------------------------------------
    # Finalize the design CSV (modified to match WAV marker format)
    # ------------------------------------------------------------------
    def finalize_design_csv(self, design_df):
        """
        Finalize the design CSV by cleaning up trial assignments and ensuring
        the WAV marker format is preserved.
        """
        # Process trial types
        baseline_mask = (design_df['trial_type'] == 'baseline')
        design_df.loc[baseline_mask, 'stimulus_type'] = None
        design_df.loc[baseline_mask, 'looming_stimulus_timestamp'] = None
        
        catch_mask = (design_df['trial_type'] == 'catch')
        design_df.loc[catch_mask, 'tactile_stimulus_timestamp'] = None
        
        practice_mask = (design_df['trial_type'] == 'practice')
        design_df.loc[practice_mask, 'stimulus_type'] = None
        design_df.loc[practice_mask, 'looming_stimulus_timestamp'] = None
        design_df.loc[practice_mask, 'tactile_stimulus_timestamp'] = None
        
        # Set expected_response based on trial type
        design_df['expected_response'] = ~design_df['trial_type'].isin(['catch', 'practice'])
        
        # Ensure we maintain the marker-based column order
        # Start with scaffold columns, then add trial-specific columns
        marker_columns = self.scaffold_columns
        trial_columns = [
            'participant_id', 'trial_number', 'trial_type', 'stimulus_type', 
            'soa_value_ms', 'jitter_ms', 'is_tactile', 'looming_stimulus_timestamp', 
            'tactile_stimulus_timestamp', 'soa_ms', 'jittered_ms', 'expected_response'
        ]
        
        # Create final column order
        all_columns = marker_columns.copy()
        for col in trial_columns:
            if col in design_df.columns and col not in all_columns:
                all_columns.append(col)
                
        # Get any remaining columns not in the predefined lists
        remaining_columns = [col for col in design_df.columns if col not in all_columns]
        all_columns.extend(remaining_columns)
        
        # Reorder and return
        final_df = design_df[all_columns].copy()
        
        self._validate_design(final_df)
        return final_df

    # ------------------------------------------------------------------
    # Design validation 
    # ------------------------------------------------------------------
    def _validate_design(self, design_df):
        """Perform comprehensive validation and counterbalancing checks on the design."""
        if not self.config['debug_mode']:
            return
            
        print("\n=== DESIGN VALIDATION ===")
        
        # Separate practice trials from experimental trials
        practice_df = design_df[design_df['trial_type'] == 'practice']
        exp_df = design_df[design_df['trial_type'] != 'practice']
        
        # 1. Overall trial count summary
        total_trials = len(design_df)
        practice_count = len(practice_df)
        experimental_count = len(exp_df)
        
        trial_counts = exp_df['trial_type'].value_counts().to_dict()
        
        print(f"TOTAL TRIALS: {total_trials} (Practice: {practice_count}, Experimental: {experimental_count})")
        print("Trial type distribution (excluding practice):")
        for trial_type, count in sorted(trial_counts.items()):
            percent = (count / experimental_count) * 100
            print(f"  {trial_type.upper()}: {count} trials ({percent:.1f}%)")
        
        # 2. Detailed breakdown by trial type
        print("\nStimulus distribution by trial type:")
        for trial_type in sorted(exp_df['trial_type'].unique()):
            if trial_type == 'baseline':
                continue
            subset = exp_df[exp_df['trial_type'] == trial_type]
            stim_counts = subset['stimulus_type'].value_counts().to_dict()
            print(f"  {trial_type.upper()}:")
            for stim, count in sorted(stim_counts.items()):
                percent = (count / len(subset)) * 100
                print(f"    {stim}: {count} trials ({percent:.1f}%)")
        
        # 3. SOA distribution checks
        print("\nSOA distribution by trial type:")
        for trial_type in sorted(exp_df['trial_type'].unique()):
            if trial_type == 'baseline':
                continue
            subset = exp_df[exp_df['trial_type'] == trial_type]
            soa_counts = subset['soa_value_ms'].value_counts().to_dict()
            print(f"  {trial_type.upper()}:")
            for soa, count in sorted(soa_counts.items()):
                percent = (count / len(subset)) * 100
                print(f"    {soa} ms: {count} trials ({percent:.1f}%)")
        
        # 4. Jitter distribution
        print("\nJitter distribution (excluding practice):")
        jitter_counts = exp_df['jitter_ms'].value_counts().to_dict()
        for jitter, count in sorted(jitter_counts.items()):
            percent = (count / experimental_count) * 100
            print(f"  {jitter} ms: {count} trials ({percent:.1f}%)")
        
        # 5. Breath phase checks for experimental trials
        print("\nBreath phase distribution (excluding practice):")
        if 'breathphase' in exp_df.columns:
            phase_counts = exp_df['breathphase'].value_counts().to_dict()
            for phase, count in sorted(phase_counts.items()):
                percent = (count / experimental_count) * 100
                print(f"  {phase}: {count} trials ({percent:.1f}%)")
        
        # 6. Cross-tabulation of trial types by breath phase
        print("\nTrial types by breath phase (COUNTERBALANCE CHECK):")
        if 'breathphase' in exp_df.columns:
            trial_by_phase = pd.crosstab(
                exp_df['trial_type'], 
                exp_df['breathphase']
            )
            print(trial_by_phase)
            
            # Calculate chi-square statistic for independence (excluding mandatory assignments)
            try:
                # Create subset without inhalation/exhalation (which are deterministically assigned)
                flex_df = exp_df[exp_df['trial_type'].isin(['baseline', 'catch'])]
                if len(flex_df) > 0:
                    flex_table = pd.crosstab(flex_df['trial_type'], flex_df['breathphase'])
                    chi2 = 0
                    row_sums = flex_table.sum(axis=1)
                    col_sums = flex_table.sum(axis=0) 
                    total = flex_table.sum().sum()
                    
                    for i in flex_table.index:
                        for j in flex_table.columns:
                            observed = flex_table.loc[i, j]
                            expected = row_sums[i] * col_sums[j] / total
                            if expected > 0:
                                chi2 += ((observed - expected) ** 2) / expected
                    
                    print(f"\nChi-square for baseline/catch distribution: {chi2:.2f}")
                    print(f"Lower values indicate better balance between breathphases.")
            except Exception as e:
                print(f"Error calculating chi-square: {e}")
        
        # 7. Consecutive trials analysis (excluding practice)
        trial_sequence = exp_df['trial_type'].tolist()
        consecutive_counts = {}
        
        if trial_sequence:
            current_type = trial_sequence[0]
            current_count = 1
            for t in trial_sequence[1:]:
                if t == current_type:
                    current_count += 1
                else:
                    key = f"{current_type}_{current_count}"
                    consecutive_counts[key] = consecutive_counts.get(key, 0) + 1
                    current_type = t
                    current_count = 1
            key = f"{current_type}_{current_count}"
            consecutive_counts[key] = consecutive_counts.get(key, 0) + 1
            
            print("\nRuns of consecutive trial types (excluding practice):")
            for key, count in sorted(consecutive_counts.items()):
                print(f"  {key}: {count}")
    
    # ------------------------------------------------------------------
    # Process a single participant: generate design, assign slots, finalize, and save CSV.
    # ------------------------------------------------------------------
    def process_participant(self, participant_id):
        try:
            if self.config['debug_mode']:
                print(f"\n=== Processing participant {participant_id} ===")
            design_df = self.generate_counterbalanced_design(participant_id)
            design_df = self.assign_breath_holds(design_df, participant_id)
            design_df = self.finalize_design_csv(design_df)
            out_path = os.path.join(self.experiment_log_dir, f"participant_{participant_id}_design.csv")
            design_df.to_csv(out_path, index=False)
            if self.config['debug_mode']:
                print(f"Saved participant {participant_id} design to {out_path}")
            return True, out_path
        except Exception as e:
            print(f"Error processing participant {participant_id}: {e}")
            traceback.print_exc()
            return False, None
    
    # ------------------------------------------------------------------
    # Run design generation for multiple participants.
    # ------------------------------------------------------------------
    def run(self, participant_ids=None, num_participants=None):
        if participant_ids is None and num_participants is None:
            num_participants = self.config.get('num_participants', 5)
        if participant_ids is None:
            participant_ids = list(range(num_participants))
        if self.config['debug_mode']:
            print("=== Starting PPSDesignGenerator with WAV Marker Integration ===")
            print(f"Marker count: {len(self.wav_markers)}")
            print(f"Repetitions: {self.config['repetitions']}")
            print(f"Practice trials: {self.config['practice_trials']}")
            print(f"Participants: {participant_ids}")
        results = {}
        for pid in participant_ids:
            success, path = self.process_participant(pid)
            results[pid] = {'success': success, 'csv_path': path}
        
        # Summary of all participants
        if self.config['debug_mode'] and len(results) > 1:
            print("\n=== CROSS-PARTICIPANT VALIDATION ===")
            all_success = all(info['success'] for info in results.values())
            print(f"All participants generated successfully: {all_success}")
        
        return results

# ----------------------------------------------------------------------
# Integrated workflow function
# ----------------------------------------------------------------------
def run_integrated_workflow(config):
    """
    Run the entire integrated workflow:
    1. Calculate required repetitions
    2. Create audio files with markers
    3. Generate participant designs
    
    Args:
        config: Configuration dictionary
        
    Returns:
        Dictionary with results
    """
    print("\n===== RUNNING INTEGRATED PPS DESIGN WORKFLOW =====")
    
    # Step 1: Initialize generator to calculate trial counts
    temp_generator = PPSDesignGenerator(config)
    
    # Step 2: Calculate repetitions needed
    repetitions_needed = temp_generator.calculate_required_repetitions()
    config['audio']['repetition_count'] = repetitions_needed
    
    print(f"\n===== AUDIO CREATION PHASE =====")
    print(f"Calculated {repetitions_needed} repetition segments needed")
    
    # Step 3: Create concatenated audio with markers
    concatenated_wav, marker_csv, extracted_markers = create_concatenated_audio_with_markers(config)
    
    print(f"\n===== DESIGN GENERATION PHASE =====")
    
    # Step 4: Initialize final generator with extracted markers
    generator = PPSDesignGenerator(config, extracted_markers)
    
    # Step 5: Generate participant designs
    participant_ids = list(range(config['num_participants']))
    results = generator.run(participant_ids)
    
    # Print summary
    print("\n===== WORKFLOW SUMMARY =====")
    print(f"Created audio files:")
    print(f"  - Concatenated WAV: {concatenated_wav}")
    print(f"  - Marker CSV: {marker_csv}")
    print(f"  - Markers extracted: {len(extracted_markers)}")
    
    success_count = sum(1 for info in results.values() if info['success'])
    print(f"Participant designs generated: {success_count}/{len(results)}")
    
    return {
        'audio_files': {
            'concatenated_wav': concatenated_wav,
            'marker_csv': marker_csv,
            'marker_count': len(extracted_markers)
        },
        'participant_designs': results
    }

# ----------------------------------------------------------------------
# MAIN FUNCTION
# ----------------------------------------------------------------------
def main():
    """Main function to run the integrated workflow."""
    # Print configuration summary
    print("\n===== INTEGRATED PPS EXPERIMENT DESIGN GENERATOR =====")
    print(f"Audio files:")
    print(f"  - Intro: {CONFIG['paths']['intro_audio']}")
    print(f"  - Repetition segment: {CONFIG['paths']['repetition_audio']}")
    print(f"  - Intro CSV: {CONFIG['paths']['intro_csv']}")
    print(f"  - Repetition CSV: {CONFIG['paths']['repetition_csv']}")
    print(f"Output directories:")
    print(f"  - Audio: {CONFIG['paths']['experiment_audio_dir']}")
    print(f"  - Design CSVs: {CONFIG['paths']['experiment_log_dir']}")
    print(f"Configuration:")
    print(f"  - Repetitions (trials per SOA per condition): {CONFIG['repetitions']}")
    print(f"  - Practice trials at beginning: {CONFIG['practice_trials']}")
    print(f"  - SOA Conditions: {CONFIG['soa_conditions_ms']}")
    print(f"  - Looming Stimuli: {LOOMING_STIMULI}")
    print(f"  - Number of Participants: {CONFIG['num_participants']}")
    
    # Run the integrated workflow
    results = run_integrated_workflow(CONFIG)
    
    # Show final summary
    print("\n===== COMPLETE =====")
    print(f"Audio file created: {results['audio_files']['concatenated_wav']}")
    print(f"Marker CSV created: {results['audio_files']['marker_csv']}")
    print(f"Participant designs created: {sum(1 for info in results['participant_designs'].values() if info['success'])}")
    print("\nProcess complete!")

if __name__ == "__main__":
    main()
```
