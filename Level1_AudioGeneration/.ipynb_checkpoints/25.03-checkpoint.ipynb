```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import soundfile as sf
import scipy.signal as signal
from datetime import datetime
from itertools import product, combinations, cycle
import warnings
import json
import math
import random
import re
from pathlib import Path
import copy
import traceback
from collections import Counter, defaultdict

# ======================================================================
# CONFIGURABLE PARAMETERS - MODIFY THESE TO CUSTOMIZE YOUR EXPERIMENT
# ======================================================================

# Default experiment configuration
EXPERIMENT_CONFIG = {
    'repetitions_per_condition': 4,  # How many times each condition is repeated
    'included_conditions': {
        'inhalation': True,  # Include inhalation trials
        'exhalation': True,  # Include exhalation trials
        'baseline': True,    # Include baseline trials
    },
    'catch_trial_percentage': 10,  # Percentage of total trials that should be catch trials
    'soa_conditions_ms': [190, 400, 700, 1000, 1500],  # SOA values in milliseconds
    'jitter_options_ms': [100, 200, 300, 400, 500],    # Jitter values in milliseconds
    'stimulus_types': ['pink', 'brown', 'white', 'blue'],  # Noise types for looming sounds
    'debug_mode': True,      # Enable for verbose output and diagnostics
    'validation': {
        'check_counterbalancing': True,   # Perform counterbalancing checks
        'generate_plots': True,           # Generate validation plots
        'save_detailed_report': True      # Save detailed validation report
    },
    'audio_extension': {
        'intro_segment_end': '02:00.2',       # Keep intro up to this point
        'repeat_segment_start': '02:00.2',    # Start repeating from here
        'repeat_segment_end': '04:56.4',      # End repeating at this point
        'preserve_ending': True,              # Keep original ending section
        'ending_duration_sec': 45,            # Duration of ending to preserve (seconds)
    },
    'paths': {
        'base_dir': r"C:\Users\cogpsy-vrlab\Documents\GitHub\BreathingSpace",
        'level0_dir': r"C:\Users\cogpsy-vrlab\Documents\GitHub\BreathingSpace\Level0_StimulusFiles",
        'experiment_log_dir': r"C:\Users\cogpsy-vrlab\Documents\GitHub\BreathingSpace\ExperimentLog",
        'audio_output_dir': r"C:\Users\cogpsy-vrlab\Documents\GitHub\BreathingSpace\ExperimentAudio",
    }
}

# File paths - These can be overridden in the EXPERIMENT_CONFIG['paths'] section
DEFAULT_PATHS = {
    'breathing_audio': "BoxBreathing.wav",
    'timestamps_csv': "BoxBreathing_timestamps.csv",
    'looming_stimuli': {
        'pink': "front_az0_FABIAN_HRIR_pink.wav",
        'brown': "front_az0_FABIAN_HRIR_brown.wav", 
        'white': "front_az0_FABIAN_HRIR_white.wav",
        'blue': "front_az0_FABIAN_HRIR_blue.wav"
    },
    'tactile_stimulus': "tactile_stimulus.wav"
}

# ======================================================================
# BREATHING CYCLE ANALYZER CLASS
# ======================================================================

class BreathingCycleAnalyzer:
    """
    Analyzes the breathing cycle timestamps to identify repeatable segments
    and calculate total cycles needed.
    """
    
    def __init__(self, timestamps_csv_path, debug_mode=False):
        """
        Initialize the analyzer with the timestamps CSV path.
        
        Args:
            timestamps_csv_path: Path to the timestamps CSV file
            debug_mode: Whether to print additional debug information
        """
        self.timestamps_path = timestamps_csv_path
        self.timestamps_df = None
        self.start_timestamps = []
        self.cycle_durations = []
        self.cycle_duration_ms = 0
        self.total_cycles = 0
        self.debug_mode = debug_mode
        self.inhale_indices = []
        self.exhale_indices = []
        
        # Load and analyze timestamps
        self._load_timestamps()
        self._analyze_cycles()
    
    def _load_timestamps(self):
        """Load timestamps from CSV file."""
        try:
            self.timestamps_df = pd.read_csv(self.timestamps_path)
            if self.debug_mode:
                print(f"Loaded {len(self.timestamps_df)} timestamps from CSV")
                print(f"Sample timestamps: {self.timestamps_df['timestamp'].head().tolist()}")
        except Exception as e:
            print(f"Error loading timestamps from {self.timestamps_path}: {e}")
            raise
    
    def _analyze_cycles(self):
        """
        Analyze timestamps to identify breathing cycles.
        Each cycle consists of:
        - Inhale
        - Hold (after inhale)
        - Exhale
        - Hold (after exhale)
        """
        if self.timestamps_df is None or self.timestamps_df.empty:
            print("Warning: No timestamps to analyze")
            return
        
        # Extract timestamps
        timestamps = self.timestamps_df['milliseconds'].values
        
        # Separate inhale and exhale timestamps
        if 'breathphase' in self.timestamps_df.columns:
            self.inhale_indices = self.timestamps_df[self.timestamps_df['breathphase'] == 'inhale'].index.tolist()
            self.exhale_indices = self.timestamps_df[self.timestamps_df['breathphase'] == 'exhale'].index.tolist()
            
            if self.debug_mode:
                print(f"Found {len(self.inhale_indices)} inhale timestamps and {len(self.exhale_indices)} exhale timestamps")
        
        # Calculate cycle information
        if len(timestamps) >= 8:  # Need at least two full cycles (4 timestamps each)
            # Calculate the duration between consecutive inhale holds (start of cycle to next cycle)
            inhale_timestamps = timestamps[self.inhale_indices] if self.inhale_indices else timestamps[::2]
            
            if len(inhale_timestamps) >= 2:
                # Each breathing cycle starts with an inhale hold
                self.start_timestamps = inhale_timestamps
                
                # Calculate durations between consecutive cycle starts
                self.cycle_durations = [inhale_timestamps[i+1] - inhale_timestamps[i] 
                                      for i in range(len(inhale_timestamps)-1)]
                
                # Calculate average cycle duration
                if self.cycle_durations:
                    self.cycle_duration_ms = sum(self.cycle_durations) / len(self.cycle_durations)
                    
                    if self.debug_mode:
                        print(f"Detected average cycle duration: {self.cycle_duration_ms/1000:.2f} seconds")
                        print(f"Individual cycle durations: {[d/1000 for d in self.cycle_durations]}")
    
    def get_cycles_for_trials(self, num_trials):
        """
        Calculate how many breathing cycles are needed for the given number of trials.
        
        Args:
            num_trials: Number of trials to accommodate
            
        Returns:
            Number of breathing cycles needed
        """
        if not self.timestamps_df.empty and len(self.inhale_indices) > 0 and len(self.exhale_indices) > 0:
            # For balanced allocation, we need equal number of inhale and exhale phases
            # Each breathing cycle has one inhale and one exhale phase
            breath_holds_per_cycle = 2  # One inhale and one exhale hold per cycle
            
            # Calculate needed cycles with buffer
            cycles_needed = math.ceil(num_trials / breath_holds_per_cycle) + 1  # Add buffer cycle
            
            if self.debug_mode:
                print(f"Need {cycles_needed} breathing cycles for {num_trials} trials "
                      f"({breath_holds_per_cycle} breath holds per cycle)")
            
            self.total_cycles = cycles_needed
            return cycles_needed
        
        # Fallback if we couldn't properly analyze the cycles
        fallback_cycles = math.ceil(num_trials / 2) + 1  # Assume 2 trials per cycle + buffer
        
        if self.debug_mode:
            print(f"Warning: Using fallback cycle calculation: {fallback_cycles} cycles for {num_trials} trials")
        
        self.total_cycles = fallback_cycles
        return fallback_cycles
    
    def get_timestamps_for_cycles(self, num_cycles):
        """
        Get the timestamps needed for the specified number of cycles.
        
        Args:
            num_cycles: Number of breathing cycles
            
        Returns:
            Array of timestamps in milliseconds
        """
        if self.timestamps_df is None or self.timestamps_df.empty:
            print("Warning: No timestamp data available")
            return []
        
        # Find how many complete cycles we have in the original data
        complete_cycles = len(self.start_timestamps) // 2  # Dividing by 2 to be conservative
        
        if self.debug_mode:
            print(f"Complete cycles in original data: {complete_cycles}")
            print(f"Cycles needed: {num_cycles}")
        
        if num_cycles <= complete_cycles:
            # We have enough cycles in the original data
            # Return timestamps for the required number of cycles (2 timestamps per cycle)
            end_idx = self.inhale_indices[num_cycles-1] if num_cycles <= len(self.inhale_indices) else len(self.timestamps_df)
            return self.timestamps_df['milliseconds'].values[:end_idx]
        else:
            # We need to extend the timestamps
            original_timestamps = self.timestamps_df['milliseconds'].values
            
            # Use the average cycle duration to extend
            if self.cycle_durations:
                avg_cycle_ms = sum(self.cycle_durations) / len(self.cycle_durations)
            else:
                # Fallback: estimate from timestamps if no cycle durations calculated
                if len(original_timestamps) >= 8:
                    avg_cycle_ms = (original_timestamps[7] - original_timestamps[3]) / 1
                else:
                    avg_cycle_ms = 8000  # Default to 8 seconds if we can't calculate
            
            # Start with original timestamps
            extended_timestamps = list(original_timestamps.copy())
            
            # Get the last timestamp
            last_ts = original_timestamps[-1]
            
            # Calculate how many more cycles we need
            additional_cycles = num_cycles - complete_cycles
            
            # Get the pattern of timestamps within a cycle
            if len(self.inhale_indices) >= 2 and len(self.exhale_indices) >= 2:
                # Use actual cycle pattern from timestamps
                cycle_start_idx = self.inhale_indices[0]
                next_cycle_start_idx = self.inhale_indices[1]
                pattern_indices = list(range(cycle_start_idx, next_cycle_start_idx))
                pattern_values = [original_timestamps[i] - original_timestamps[cycle_start_idx] for i in pattern_indices]
                
                if self.debug_mode:
                    print(f"Detected timestamp pattern within cycle: {pattern_values}")
            else:
                # Fallback pattern (evenly spaced timestamps)
                pattern_values = [0, avg_cycle_ms/4, avg_cycle_ms/2, avg_cycle_ms*3/4]
                
                if self.debug_mode:
                    print(f"Using fallback timestamp pattern: {pattern_values}")
            
            # Extend by adding cycles
            for c in range(additional_cycles):
                cycle_base = last_ts + c * avg_cycle_ms
                
                # Add timestamps for this cycle based on the pattern
                for offset in pattern_values:
                    extended_timestamps.append(cycle_base + offset)
            
            if self.debug_mode:
                print(f"Extended timestamps: Original: {len(original_timestamps)}, "
                      f"Extended: {len(extended_timestamps)}")
            
            return np.array(extended_timestamps)
    
    def get_breathphase_for_timestamp(self, timestamp_ms):
        """
        Determine the breathing phase for a given timestamp.
        
        Args:
            timestamp_ms: Timestamp in milliseconds
            
        Returns:
            'inhale', 'exhale', or None if can't determine
        """
        if self.timestamps_df is None or self.timestamps_df.empty:
            return None
        
        # Find the closest timestamp in the data
        timestamps = self.timestamps_df['milliseconds'].values
        idx = np.abs(timestamps - timestamp_ms).argmin()
        
        # Check if timestamp is close enough to be associated with this breath phase
        closest_diff = abs(timestamps[idx] - timestamp_ms)
        if closest_diff > 1000:  # More than 1 second difference
            return None
        
        # Get the breath phase from the closest timestamp
        if 'breathphase' in self.timestamps_df.columns:
            return self.timestamps_df.iloc[idx]['breathphase']
        
        # Fallback: try to infer from pattern (odd indices are exhale, even are inhale)
        return 'inhale' if idx % 2 == 0 else 'exhale'
    
    def get_segmentation_indices(self, intro_end, repeat_start, repeat_end):
        """
        Get the indices corresponding to the segmentation points.
        
        Args:
            intro_end: Timestamp string for end of intro segment
            repeat_start: Timestamp string for start of repeatable segment
            repeat_end: Timestamp string for end of repeatable segment
            
        Returns:
            Tuple of (intro_end_idx, repeat_start_idx, repeat_end_idx)
        """
        if self.timestamps_df is None or self.timestamps_df.empty:
            return None, None, None
        
        # Find indices in timestamps dataframe
        intro_end_idx = None
        repeat_start_idx = None
        repeat_end_idx = None
        
        # Try to find exact matches first
        if 'timestamp' in self.timestamps_df.columns:
            intro_end_matches = self.timestamps_df[self.timestamps_df['timestamp'] == intro_end].index
            repeat_start_matches = self.timestamps_df[self.timestamps_df['timestamp'] == repeat_start].index
            repeat_end_matches = self.timestamps_df[self.timestamps_df['timestamp'] == repeat_end].index
            
            if len(intro_end_matches) > 0:
                intro_end_idx = intro_end_matches[0]
            if len(repeat_start_matches) > 0:
                repeat_start_idx = repeat_start_matches[0]
            if len(repeat_end_matches) > 0:
                repeat_end_idx = repeat_end_matches[0]
        
        # If we couldn't find exact matches, try approximation
        if intro_end_idx is None or repeat_start_idx is None or repeat_end_idx is None:
            # Convert timestamp strings to seconds
            intro_end_sec = self._timestamp_to_seconds(intro_end)
            repeat_start_sec = self._timestamp_to_seconds(repeat_start)
            repeat_end_sec = self._timestamp_to_seconds(repeat_end)
            
            # Add time_seconds column if not present
            if 'time_seconds' not in self.timestamps_df.columns:
                self.timestamps_df['time_seconds'] = self.timestamps_df['milliseconds'] / 1000
            
            # Find closest matches
            intro_end_idx = (self.timestamps_df['time_seconds'] - intro_end_sec).abs().idxmin()
            repeat_start_idx = (self.timestamps_df['time_seconds'] - repeat_start_sec).abs().idxmin()
            repeat_end_idx = (self.timestamps_df['time_seconds'] - repeat_end_sec).abs().idxmin()
        
        return intro_end_idx, repeat_start_idx, repeat_end_idx
    
    def _timestamp_to_seconds(self, timestamp_str):
        """
        Convert a timestamp string (MM:SS.F) to seconds.
        
        Args:
            timestamp_str: Timestamp string in format MM:SS.F
            
        Returns:
            Time in seconds
        """
        match = re.match(r'(\d+):(\d+\.\d+)', timestamp_str)
        if match:
            minutes, seconds = match.groups()
            return int(minutes) * 60 + float(seconds)
        return 0


# ======================================================================
# ENHANCED PPS AUDIO GENERATOR
# ======================================================================

class EnhancedPPSAudioGenerator:
    """
    Generates audio files for Peripersonal Space (PPS) experiments with
    precise control over trial repetitions, sophisticated breathing cycle analysis,
    and comprehensive counterbalancing validation.
    """
    
    def __init__(self, config=None):
        """
        Initialize the audio generator with configuration parameters.
        
        Parameters:
        -----------
        config : dict, optional
            Configuration dictionary with experiment parameters
        """
        # Use default config and update with provided config
        self.config = copy.deepcopy(EXPERIMENT_CONFIG)
        if config:
            # Recursively update nested dictionaries
            self._update_nested_dict(self.config, config)
        
        # Set up paths
        self.setup_paths()
        
        # Initialize counters and trackers
        self.trial_counts = self._calculate_trial_counts()
        self.conditions = [cond for cond, include in self.config['included_conditions'].items() if include]
        
        # Initialize empty containers for data
        self.timestamps_df = None
        self.breathing_analyzer = None
        self.extended_timestamps_df = None
        self.extended_audio = None
        self.extended_audio_path = None
        self.sample_rate = None
        self.looming_stimuli = {}
        self.tactile_stimulus = None
        
        if self.config['debug_mode']:
            print(f"Initialized EnhancedPPSAudioGenerator with:")
            print(f"  - Repetitions: {self.config['repetitions_per_condition']}")
            print(f"  - Conditions: {self.conditions}")
            print(f"  - SOA values: {self.config['soa_conditions_ms']}")
            print(f"  - Total trials per participant: {sum(self.trial_counts.values())}")
    
    def _update_nested_dict(self, d, u):
        """Recursively update nested dictionary."""
        for k, v in u.items():
            if isinstance(v, dict) and k in d and isinstance(d[k], dict):
                self._update_nested_dict(d[k], v)
            else:
                d[k] = v
    
    def _calculate_trial_counts(self):
        """
        Calculate trial counts based on the configuration.
        
        Returns:
            Dictionary with trial counts for each condition
        """
        # Get parameters
        reps = self.config['repetitions_per_condition']
        soa_count = len(self.config['soa_conditions_ms'])
        catch_percentage = self.config['catch_trial_percentage']
        
        # Calculate trials for each condition
        trial_counts = {}
        for condition, include in self.config['included_conditions'].items():
            if include:
                # For each condition, we need (# stimulus types) × (# SOA values) × (repetitions) trials
                trial_counts[condition] = len(self.config['stimulus_types']) * soa_count * reps
        
        # Calculate catch trials
        regular_trials = sum(trial_counts.values())
        catch_trials = int(np.ceil(regular_trials * catch_percentage / 100))
        trial_counts['catch'] = catch_trials
        
        if self.config['debug_mode']:
            print(f"Trial count calculation:")
            for cond, count in trial_counts.items():
                print(f"  {cond}: {count} trials")
            print(f"  Total: {sum(trial_counts.values())} trials")
        
        return trial_counts
    
    def setup_paths(self):
        """
        Set up file paths for input and output files.
        """
        # Get base paths from config
        base_dir = self.config['paths']['base_dir']
        level0_dir = self.config['paths']['level0_dir']
        self.experiment_log_dir = self.config['paths']['experiment_log_dir']
        self.audio_output_dir = self.config['paths']['audio_output_dir']
        
        # Input files
        self.breathing_audio_path = os.path.join(level0_dir, DEFAULT_PATHS['breathing_audio'])
        self.timestamps_csv_path = os.path.join(level0_dir, DEFAULT_PATHS['timestamps_csv'])
        
        # Stimulus files
        self.looming_stimuli_paths = {}
        for stim_type in self.config['stimulus_types']:
            if stim_type in DEFAULT_PATHS['looming_stimuli']:
                self.looming_stimuli_paths[stim_type] = os.path.join(
                    level0_dir, DEFAULT_PATHS['looming_stimuli'][stim_type]
                )
        
        self.tactile_stimulus_path = os.path.join(level0_dir, DEFAULT_PATHS['tactile_stimulus'])
        
        # Create output directories if they don't exist
        os.makedirs(self.experiment_log_dir, exist_ok=True)
        os.makedirs(self.audio_output_dir, exist_ok=True)
        
        # Create validation directory if needed
        if self.config['validation']['generate_plots']:
            self.validation_plots_dir = os.path.join(self.experiment_log_dir, 'validation_plots')
            os.makedirs(self.validation_plots_dir, exist_ok=True)
        
        if self.config['debug_mode']:
            print(f"Set up paths:")
            print(f"  - Breathing audio: {self.breathing_audio_path}")
            print(f"  - Timestamps CSV: {self.timestamps_csv_path}")
            print(f"  - Looming stimuli: {len(self.looming_stimuli_paths)} files")
            print(f"  - Tactile stimulus: {self.tactile_stimulus_path}")
            print(f"  - Output directories created")
        
        # Verify essential files exist
        missing_files = []
        if not os.path.exists(self.breathing_audio_path):
            missing_files.append(self.breathing_audio_path)
        if not os.path.exists(self.timestamps_csv_path):
            missing_files.append(self.timestamps_csv_path)
        for path in self.looming_stimuli_paths.values():
            if not os.path.exists(path):
                missing_files.append(path)
        if not os.path.exists(self.tactile_stimulus_path):
            missing_files.append(self.tactile_stimulus_path)
        
        if missing_files:
            print(f"Warning: Missing input files:")
            for file in missing_files:
                print(f"  - {file}")
            print("Please check your file paths configuration.")
    
    def initialize_breathing_analyzer(self):
        """
        Initialize the breathing cycle analyzer.
        
        Returns:
            BreathingCycleAnalyzer instance
        """
        if self.breathing_analyzer is None:
            try:
                self.breathing_analyzer = BreathingCycleAnalyzer(
                    self.timestamps_csv_path, 
                    debug_mode=self.config['debug_mode']
                )
            except Exception as e:
                print(f"Error initializing breathing analyzer: {e}")
                traceback.print_exc()
        
        return self.breathing_analyzer
    
    def parse_timestamps(self):
        """
        Parse the box breathing timestamps CSV.
        
        Returns:
        --------
        pd.DataFrame
            Processed timestamps with additional columns
        """
        if self.timestamps_df is not None:
            return self.timestamps_df
        
        if self.breathing_analyzer is not None:
            self.timestamps_df = self.breathing_analyzer.timestamps_df
            return self.timestamps_df
        
        try:
            # Read the CSV file
            df = pd.read_csv(self.timestamps_csv_path)
            
            # Ensure we have the expected columns
            required_columns = ['timestamp', 'milliseconds', 'sample_index']
            for col in required_columns:
                if col not in df.columns:
                    raise ValueError(f"Required column '{col}' not found in timestamps CSV")
            
            # Add breath phase if not present
            if 'breathphase' not in df.columns:
                # Try to infer from pattern or description
                if 'description' in df.columns:
                    # Look for keywords in description
                    df['breathphase'] = df['description'].apply(
                        lambda x: 'inhale' if 'inhale' in str(x).lower() else 'exhale' if 'exhale' in str(x).lower() else None
                    )
                else:
                    # Infer from pattern (even indices are inhale, odd are exhale)
                    df['breathphase'] = ['inhale' if i % 2 == 0 else 'exhale' for i in range(len(df))]
            
            # Convert timestamp strings to seconds for easier processing
            def timestamp_to_seconds(ts):
                match = re.match(r'(\d+):(\d+\.\d+)', ts)
                if match:
                    minutes, seconds = match.groups()
                    return int(minutes) * 60 + float(seconds)
                return 0
            
            df['time_seconds'] = df['timestamp'].apply(timestamp_to_seconds)
            
            # Store the processed dataframe
            self.timestamps_df = df
            
            if self.config['debug_mode']:
                inhale_count = sum(df['breathphase'] == 'inhale')
                exhale_count = sum(df['breathphase'] == 'exhale')
                print(f"Parsed timestamps: {len(df)} total entries")
                print(f"Breath phases: {inhale_count} inhale, {exhale_count} exhale")
            
            return df
            
        except Exception as e:
            print(f"Error parsing timestamps: {e}")
            traceback.print_exc()
            return pd.DataFrame()  # Return empty dataframe on error
    
    def analyze_audio_extension_needs(self):
        """
        Analyze whether audio extension is needed based on trial count.
        
        Returns:
        --------
        tuple
            (extension_needed, required_breath_holds, available_breath_holds)
        """
        # Ensure we have a breathing analyzer
        self.initialize_breathing_analyzer()
        
        # Calculate required trials
        total_trials = sum(self.trial_counts.values())
        
        # We need breath holds for all trials except catch trials
        required_breath_holds = total_trials - self.trial_counts.get('catch', 0)
        
        # Get available breath holds from timestamps
        timestamps_df = self.parse_timestamps()
        
        # Get segmentation indices
        intro_end = self.config['audio_extension']['intro_segment_end']
        repeat_start = self.config['audio_extension']['repeat_segment_start']
        repeat_end = self.config['audio_extension']['repeat_segment_end']
        
        intro_end_idx, repeat_start_idx, repeat_end_idx = self.breathing_analyzer.get_segmentation_indices(
            intro_end, repeat_start, repeat_end
        )
        
        if intro_end_idx is None or repeat_start_idx is None or repeat_end_idx is None:
            print("Warning: Could not find segmentation indices in timestamps")
            # Use default values (reserve first 10% of timestamps)
            total_timestamps = len(timestamps_df)
            intro_end_idx = int(total_timestamps * 0.1)
            repeat_start_idx = intro_end_idx
            repeat_end_idx = int(total_timestamps * 0.8)  # Reserve last 20% for ending
        
        # Calculate available usable breath holds (after intro, before ending)
        available_breath_holds = len(timestamps_df) - (intro_end_idx + 1)
        
        # Check if extension is needed
        extension_needed = required_breath_holds > available_breath_holds
        
        if self.config['debug_mode']:
            print(f"Audio extension analysis:")
            print(f"  Required breath holds: {required_breath_holds}")
            print(f"  Available breath holds: {available_breath_holds}")
            print(f"  Extension needed: {extension_needed}")
            print(f"  Segmentation indices: intro_end={intro_end_idx}, "
                  f"repeat_start={repeat_start_idx}, repeat_end={repeat_end_idx}")
        
        return (
            extension_needed, 
            required_breath_holds, 
            available_breath_holds, 
            (intro_end_idx, repeat_start_idx, repeat_end_idx)
        )
    
    def extend_breathing_audio(self):
        """
        Extend the box breathing audio to accommodate all trials.
        
        Creates an extended audio file by duplicating segments of the
        original audio according to the configuration parameters.
        
        Returns:
        --------
        tuple
            (extended_audio_path, extended_timestamps_df)
        """
        # Ensure we have a breathing analyzer
        self.initialize_breathing_analyzer()
        
        # Check if extension is needed
        extension_info = self.analyze_audio_extension_needs()
        extension_needed, required_holds, available_holds, segment_indices = extension_info
        intro_end_idx, repeat_start_idx, repeat_end_idx = segment_indices
        
        if not extension_needed:
            if self.config['debug_mode']:
                print("Audio extension not needed, using original audio file.")
            return self.breathing_audio_path, self.parse_timestamps()
        
        # Load original audio
        original_audio, sample_rate = sf.read(self.breathing_audio_path)
        self.sample_rate = sample_rate
        
        # Parse timestamps
        timestamps_df = self.parse_timestamps()
        
        # Get sample indices for these boundaries
        intro_end_sample = int(timestamps_df.iloc[intro_end_idx]['sample_index'])
        repeat_start_sample = int(timestamps_df.iloc[repeat_start_idx]['sample_index'])
        repeat_end_sample = int(timestamps_df.iloc[repeat_end_idx]['sample_index'])
        
        # Calculate segment size in breath holds
        segment_size = repeat_end_idx - repeat_start_idx + 1
        segment_samples = repeat_end_sample - repeat_start_sample
        
        # Calculate how many additional segments we need
        additional_holds_needed = required_holds - available_holds
        additional_segments_needed = int(np.ceil(additional_holds_needed / segment_size))
        
        if self.config['debug_mode']:
            print(f"Audio extension details:")
            print(f"  Segment size: {segment_size} breath holds ({segment_samples} samples)")
            print(f"  Additional holds needed: {additional_holds_needed}")
            print(f"  Additional segments needed: {additional_segments_needed}")
        
        # Create extended audio
        # Keep everything up to the intro_end
        extended_audio = original_audio[:intro_end_sample].copy()
        
        # Add the number of segments needed
        for i in range(additional_segments_needed + 1):  # +1 to keep the original segment
            segment = original_audio[repeat_start_sample:repeat_end_sample].copy()
            if len(extended_audio.shape) > 1 and len(segment.shape) > 1:
                # Both are stereo
                extended_audio = np.vstack([extended_audio, segment])
            else:
                # At least one is mono
                if len(extended_audio.shape) > 1:
                    # Extended is stereo, segment is mono
                    segment_stereo = np.column_stack([segment, segment])
                    extended_audio = np.vstack([extended_audio, segment_stereo])
                elif len(segment.shape) > 1:
                    # Extended is mono, segment is stereo
                    extended_mono = np.column_stack([extended_audio, extended_audio])
                    segment_mono = np.mean(segment, axis=1)
                    extended_audio = np.vstack([extended_mono, segment_mono])
                else:
                    # Both are mono
                    extended_audio = np.concatenate([extended_audio, segment])
        
        # Add the remaining part if we want to preserve the ending
        if self.config['audio_extension']['preserve_ending']:
            # Either add the original ending from repeat_end_sample to end
            # or use a fixed duration ending
            ending_duration_sec = self.config['audio_extension']['ending_duration_sec']
            ending_samples = min(
                len(original_audio) - repeat_end_sample,
                int(ending_duration_sec * sample_rate)
            )
            
            # Get ending segment
            ending_segment = original_audio[repeat_end_sample:repeat_end_sample+ending_samples].copy()
            
            # Add ending segment
            if len(extended_audio.shape) > 1 and len(ending_segment.shape) > 1:
                # Both are stereo
                extended_audio = np.vstack([extended_audio, ending_segment])
            else:
                # At least one is mono
                if len(extended_audio.shape) > 1:
                    # Extended is stereo, ending is mono
                    ending_stereo = np.column_stack([ending_segment, ending_segment])
                    extended_audio = np.vstack([extended_audio, ending_stereo])
                elif len(ending_segment.shape) > 1:
                    # Extended is mono, ending is stereo
                    extended_mono = np.column_stack([extended_audio, extended_audio])
                    ending_mono = np.mean(ending_segment, axis=1)
                    extended_audio = np.vstack([extended_mono, ending_mono])
                else:
                    # Both are mono
                    extended_audio = np.concatenate([extended_audio, ending_segment])
        
        # Create extended timestamps
        # Keep original timestamps up to intro_end
        extended_timestamps = timestamps_df.iloc[:intro_end_idx+1].copy()
        
        # Add the extended segments
        time_offset = 0
        sample_offset = 0
        milliseconds_offset = 0
        
        for i in range(additional_segments_needed + 1):  # +1 to keep the original segment
            segment_timestamps = timestamps_df.iloc[repeat_start_idx:repeat_end_idx+1].copy()
            
            if i > 0:  # Skip offset for first segment (it's the original one)
                # Calculate offsets based on the segment duration
                segment_duration = (timestamps_df.iloc[repeat_end_idx]['time_seconds'] - 
                                  timestamps_df.iloc[repeat_start_idx]['time_seconds'])
                segment_milliseconds = (timestamps_df.iloc[repeat_end_idx]['milliseconds'] - 
                                      timestamps_df.iloc[repeat_start_idx]['milliseconds'])
                
                time_offset += segment_duration
                milliseconds_offset += segment_milliseconds
                sample_offset += segment_samples
                
                # Apply offsets to this segment
                segment_timestamps['time_seconds'] += time_offset
                segment_timestamps['milliseconds'] += milliseconds_offset
                segment_timestamps['sample_index'] += sample_offset
                
                # Update timestamp strings
                for idx, row in segment_timestamps.iterrows():
                    minutes = int(row['time_seconds'] // 60)
                    seconds = row['time_seconds'] % 60
                    segment_timestamps.at[idx, 'timestamp'] = f"{minutes:02d}:{seconds:.1f}"
            
            # Append this segment to extended timestamps
            extended_timestamps = pd.concat([extended_timestamps, segment_timestamps])
        
        # Add ending timestamps if preserving ending
        if self.config['audio_extension']['preserve_ending']:
            remaining_timestamps = timestamps_df.iloc[repeat_end_idx+1:].copy()
            
            if not remaining_timestamps.empty:
                # Apply final offsets to remaining
                remaining_timestamps['time_seconds'] += time_offset
                remaining_timestamps['milliseconds'] += milliseconds_offset
                remaining_timestamps['sample_index'] += sample_offset
                
                # Update timestamp strings for remaining
                for idx, row in remaining_timestamps.iterrows():
                    minutes = int(row['time_seconds'] // 60)
                    seconds = row['time_seconds'] % 60
                    remaining_timestamps.at[idx, 'timestamp'] = f"{minutes:02d}:{seconds:.1f}"
                
                # Append remaining
                extended_timestamps = pd.concat([extended_timestamps, remaining_timestamps])
        
        # Reset index
        extended_timestamps = extended_timestamps.reset_index(drop=True)
        
        # Save extended audio file
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        extended_audio_path = os.path.join(
            self.audio_output_dir, 
            f"extended_boxbreathing_{timestamp}.wav"
        )
        sf.write(extended_audio_path, extended_audio, sample_rate)
        
        # Save extended timestamps
        extended_timestamps_path = os.path.join(
            self.experiment_log_dir, 
            f"extended_boxbreathing_timestamps_{timestamp}.csv"
        )
        extended_timestamps.to_csv(extended_timestamps_path, index=False)
        
        if self.config['debug_mode']:
            print(f"Extended audio created:")
            print(f"  Original length: {len(original_audio)/sample_rate:.2f} seconds")
            print(f"  Extended length: {len(extended_audio)/sample_rate:.2f} seconds")
            print(f"  Extended audio saved to: {extended_audio_path}")
            print(f"  Extended timestamps saved to: {extended_timestamps_path}")
            print(f"  Original timestamps: {len(timestamps_df)}, Extended: {len(extended_timestamps)}")
        
        # Store for later use
        self.extended_audio = extended_audio
        self.extended_audio_path = extended_audio_path
        self.extended_timestamps_df = extended_timestamps
        
        return extended_audio_path, extended_timestamps
    
    def load_stimuli(self):
        """
        Load all stimulus files.
        
        Returns:
        --------
        tuple
            (looming_stimuli_dict, tactile_stimulus)
        """
        if self.looming_stimuli and self.tactile_stimulus is not None:
            return self.looming_stimuli, self.tactile_stimulus
        
        # Load looming stimuli
        looming_stimuli = {}
        for stim_type, path in self.looming_stimuli_paths.items():
            try:
                data, sr = sf.read(path)
                if self.sample_rate is None:
                    self.sample_rate = sr
                elif sr != self.sample_rate:
                    print(f"Warning: Sample rate mismatch for {stim_type}: {sr} vs {self.sample_rate}")
                looming_stimuli[stim_type] = data
                if self.config['debug_mode']:
                    print(f"Loaded {stim_type} looming stimulus: {len(data)} samples, shape: {data.shape}")
            except Exception as e:
                print(f"Error loading {stim_type} stimulus: {e}")
        
        # Load tactile stimulus
        try:
            tactile_data, sr = sf.read(self.tactile_stimulus_path)
            if self.sample_rate is None:
                self.sample_rate = sr
            elif sr != self.sample_rate:
                print(f"Warning: Sample rate mismatch for tactile: {sr} vs {self.sample_rate}")
            self.tactile_stimulus = tactile_data
            if self.config['debug_mode']:
                print(f"Loaded tactile stimulus: {len(tactile_data)} samples, shape: {tactile_data.shape}")
        except Exception as e:
            print(f"Error loading tactile stimulus: {e}")
            self.tactile_stimulus = None
        
        # Store for later use
        self.looming_stimuli = looming_stimuli
        
        return looming_stimuli, self.tactile_stimulus
    
    def generate_counterbalanced_design(self, participant_id):
        """
        Generate a counterbalanced design for a participant.
        
        Creates a balanced distribution of trial types, SOA conditions,
        and stimulus types according to the configuration parameters.
        
        Parameters:
        -----------
        participant_id : int
            Participant identifier
        
        Returns:
        --------
        pd.DataFrame
            Design dataframe with all trial parameters
        """
        # Get parameters
        soa_conditions = self.config['soa_conditions_ms']
        jitter_options = self.config['jitter_options_ms']
        stimulus_types = self.config['stimulus_types']
        
        # Create lists to hold all trials
        all_trials = []
        
        # Generate regular trials for each condition
        for condition, count in self.trial_counts.items():
            # Skip catch trials (handled separately) or conditions with zero trials
            if condition == 'catch' or count == 0:
                continue
            
            # Create balanced combinations of stimulus type and SOA
            stim_soa_combinations = list(product(stimulus_types, soa_conditions))
            
            # Ensure proper repetition
            reps_needed = count // len(stim_soa_combinations)
            remainder = count % len(stim_soa_combinations)
            
            # Use full repetitions
            full_combinations = stim_soa_combinations * reps_needed
            
            # Add remainder trials (randomly selected to maintain balance)
            remainder_combinations = random.sample(stim_soa_combinations, remainder)
            
            # Combine all combinations
            all_combinations = full_combinations + remainder_combinations
            
            # Shuffle to randomize order
            random.shuffle(all_combinations)
            
            # Create trial entries
            for i, (stim_type, soa) in enumerate(all_combinations):
                all_trials.append({
                    'participant_id': participant_id,
                    'trial_number': len(all_trials) + 1,
                    'stimulus_type': stim_type,
                    'trial_type': condition,
                    'soa_condition_idx': soa_conditions.index(soa),
                    'soa_value_ms': soa,
                    'jitter_ms': random.choice(jitter_options),
                    'is_tactile': True  # Regular trials have tactile component
                })
        
        # Add catch trials
        catch_count = self.trial_counts.get('catch', 0)
        if catch_count > 0:
            # Create balanced catch trials (no tactile stimulus)
            catch_stim_combinations = list(product(stimulus_types, soa_conditions))
            
            # Ensure enough combinations
            while len(catch_stim_combinations) < catch_count:
                catch_stim_combinations.extend(catch_stim_combinations)
            
            # Select random subset if too many
            catch_combinations = random.sample(catch_stim_combinations, catch_count)
            
            # Create catch trial entries
            for i, (stim_type, soa) in enumerate(catch_combinations):
                all_trials.append({
                    'participant_id': participant_id,
                    'trial_number': len(all_trials) + 1,
                    'stimulus_type': stim_type,
                    'trial_type': 'catch',
                    'soa_condition_idx': soa_conditions.index(soa),
                    'soa_value_ms': soa,
                    'jitter_ms': random.choice(jitter_options),
                    'is_tactile': False  # Catch trials have no tactile component
                })
        
        # Create DataFrame
        design_df = pd.DataFrame(all_trials)
        
        # Shuffle the design to randomize trial order
        # But use a consistent seed based on participant ID for reproducibility
        design_df = design_df.sample(frac=1, random_state=participant_id*100).reset_index(drop=True)
        
        # Update trial numbers after shuffling
        design_df['trial_number'] = range(1, len(design_df) + 1)
        
        if self.config['debug_mode']:
            print(f"Generated counterbalanced design for participant {participant_id}:")
            print(f"  Total trials: {len(design_df)}")
            print(f"  Trial types: {design_df['trial_type'].value_counts().to_dict()}")
            print(f"  Stimulus types: {design_df['stimulus_type'].value_counts().to_dict()}")
            print(f"  SOA values: {design_df['soa_value_ms'].value_counts().to_dict()}")
        
        return design_df
    
    def assign_trials_to_breath_holds(self, design_df, timestamps_df):
        """
        Assign trials to specific breath holds while maintaining balance.
        
        Parameters:
        -----------
        design_df : pd.DataFrame
            Design dataframe with trial parameters
        timestamps_df : pd.DataFrame
            Timestamps dataframe (original or extended)
        
        Returns:
        --------
        pd.DataFrame
            Design dataframe with breath hold assignments
        """
        # Ensure we have a breathing analyzer
        if self.breathing_analyzer is None:
            self.initialize_breathing_analyzer()
        
        # Get audio extension parameters
        intro_end = self.config['audio_extension']['intro_segment_end']
        
        # Get segmentation indices
        indices = self.breathing_analyzer.get_segmentation_indices(
            intro_end, intro_end, intro_end  # We only need intro_end
        )
        intro_end_idx = indices[0]
        
        if intro_end_idx is None:
            # Fallback to a percentage of total timestamps
            intro_end_idx = int(len(timestamps_df) * 0.1)
            print(f"Warning: Using fallback intro_end_idx = {intro_end_idx}")
        
        # Separate inhale and exhale timestamps
        # Skip first few timestamps (introduction period)
        usable_timestamps = timestamps_df.iloc[intro_end_idx+1:].copy()
        inhale_timestamps = usable_timestamps[usable_timestamps['breathphase'] == 'inhale']
        exhale_timestamps = usable_timestamps[usable_timestamps['breathphase'] == 'exhale']
        
        if self.config['debug_mode']:
            print(f"Assigning trials to breath holds:")
            print(f"  Available inhale holds: {len(inhale_timestamps)}")
            print(f"  Available exhale holds: {len(exhale_timestamps)}")
        
        # Create copies for assignment
        inhale_indices = inhale_timestamps.index.tolist()
        exhale_indices = exhale_timestamps.index.tolist()
        
        # Random shuffle both lists
        random.shuffle(inhale_indices)
        random.shuffle(exhale_indices)
        
        # Assign trials to breath holds
        trial_assignments = {}
        
        # Group trials by condition
        for condition in set(design_df['trial_type']):
            condition_trials = design_df[design_df['trial_type'] == condition]
            condition_count = len(condition_trials)
            
            # Calculate how many trials of each breath phase we need
            inhale_count = condition_count // 2
            exhale_count = condition_count - inhale_count
            
            if self.config['debug_mode']:
                print(f"  Condition '{condition}': {condition_count} trials - {inhale_count} inhale, {exhale_count} exhale")
            
            if inhale_count > len(inhale_indices) or exhale_count > len(exhale_indices):
                print(f"Warning: Not enough breath holds for condition '{condition}'")
                print(f"  Need {inhale_count} inhale holds, have {len(inhale_indices)}")
                print(f"  Need {exhale_count} exhale holds, have {len(exhale_indices)}")
                print("  Some trials may not be assigned to breath holds")
                
                # Adjust to use what we have
                inhale_count = min(inhale_count, len(inhale_indices))
                exhale_count = min(exhale_count, len(exhale_indices))
            
            # Assign to inhale holds
            for idx, trial_idx in zip(inhale_indices[:inhale_count], condition_trials.index[:inhale_count]):
                timestamp_row = timestamps_df.loc[idx]
                trial_assignments[trial_idx] = {
                    'timestamp_idx': idx,
                    'breathphase': 'inhale',
                    'timestamp': timestamp_row['timestamp'],
                    'milliseconds': timestamp_row['milliseconds'],
                    'sample_index': timestamp_row['sample_index'],
                    'time_seconds': timestamp_row['time_seconds'] if 'time_seconds' in timestamp_row else timestamp_row['milliseconds'] / 1000
                }
            
            # Assign to exhale holds
            for idx, trial_idx in zip(exhale_indices[:exhale_count], condition_trials.index[inhale_count:inhale_count+exhale_count]):
                timestamp_row = timestamps_df.loc[idx]
                trial_assignments[trial_idx] = {
                    'timestamp_idx': idx,
                    'breathphase': 'exhale',
                    'timestamp': timestamp_row['timestamp'],
                    'milliseconds': timestamp_row['milliseconds'],
                    'sample_index': timestamp_row['sample_index'],
                    'time_seconds': timestamp_row['time_seconds'] if 'time_seconds' in timestamp_row else timestamp_row['milliseconds'] / 1000
                }
            
            # Remove assigned indices to prevent reuse
            inhale_indices = inhale_indices[inhale_count:]
            exhale_indices = exhale_indices[exhale_count:]
        
        # Add assignment info to design dataframe
        design_df['breathphase'] = design_df.index.map(lambda idx: trial_assignments.get(idx, {}).get('breathphase'))
        design_df['timestamp'] = design_df.index.map(lambda idx: trial_assignments.get(idx, {}).get('timestamp'))
        design_df['milliseconds'] = design_df.index.map(lambda idx: trial_assignments.get(idx, {}).get('milliseconds'))
        design_df['sample_index'] = design_df.index.map(lambda idx: trial_assignments.get(idx, {}).get('sample_index'))
        design_df['time_seconds'] = design_df.index.map(lambda idx: trial_assignments.get(idx, {}).get('time_seconds'))
        
        # Check if any trials were not assigned
        unassigned = design_df[design_df['breathphase'].isna()]
        if not unassigned.empty:
            print(f"Warning: {len(unassigned)} trials could not be assigned to breath holds")
            
            # Fallback assignment - assign to random timestamps
            available_timestamps = timestamps_df.loc[
                ~timestamps_df.index.isin([a['timestamp_idx'] for a in trial_assignments.values() if 'timestamp_idx' in a])
            ]
            
            if not available_timestamps.empty:
                print(f"  Using fallback assignment with {len(available_timestamps)} available timestamps")
                for idx in unassigned.index:
                    if available_timestamps.empty:
                        break
                    
                    # Get a random timestamp
                    ts_idx = random.choice(available_timestamps.index)
                    timestamp_row = available_timestamps.loc[ts_idx]
                    
                    # Assign to this trial
                    design_df.at[idx, 'breathphase'] = timestamp_row.get('breathphase', 'unknown')
                    design_df.at[idx, 'timestamp'] = timestamp_row['timestamp']
                    design_df.at[idx, 'milliseconds'] = timestamp_row['milliseconds']
                    design_df.at[idx, 'sample_index'] = timestamp_row['sample_index']
                    design_df.at[idx, 'time_seconds'] = timestamp_row.get('time_seconds', timestamp_row['milliseconds'] / 1000)
                    
                    # Remove this timestamp from available ones
                    available_timestamps = available_timestamps.drop(ts_idx)
            else:
                print("  No available timestamps for fallback assignment")
        
        # Calculate timestamps with jitter
        design_df['jittered_ms'] = design_df.apply(
            lambda row: row['milliseconds'] + row['jitter_ms'] if pd.notnull(row['milliseconds']) else None,
            axis=1
        )
        
        # Calculate timestamps with SOA for tactile stimulus (only for non-catch trials)
        design_df['soa_ms'] = design_df.apply(
            lambda row: row['jittered_ms'] + row['soa_value_ms'] if pd.notnull(row['jittered_ms']) and row['trial_type'] != 'catch' else None, 
            axis=1
        )
        
        # Format timestamps for output
        design_df['timestamp_original'] = design_df.apply(
            lambda row: f"{int(row['milliseconds'] // 60000):02}:{(row['milliseconds'] % 60000) / 1000:.1f}" if pd.notnull(row['milliseconds']) else None,
            axis=1
        )
        
        design_df['timestamp_after_jitter'] = design_df.apply(
            lambda row: f"{int(row['jittered_ms'] // 60000):02}:{(row['jittered_ms'] % 60000) / 1000:.1f}" if pd.notnull(row['jittered_ms']) else None,
            axis=1
        )
        
        design_df['timestamp_with_soa'] = design_df.apply(
            lambda row: f"{int(row['soa_ms'] // 60000):02}:{(row['soa_ms'] % 60000) / 1000:.1f}" if pd.notnull(row['soa_ms']) else None,
            axis=1
        )
        
        # Convert milliseconds to sample indices
        sample_rate = self.sample_rate if self.sample_rate else 48000
        design_df['base_ms'] = design_df['milliseconds']
        design_df['jittered_sample'] = design_df.apply(
            lambda row: int(row['jittered_ms'] / 1000 * sample_rate) if pd.notnull(row['jittered_ms']) else None,
            axis=1
        )
        design_df['soa_sample'] = design_df.apply(
            lambda row: int(row['soa_ms'] / 1000 * sample_rate) if pd.notnull(row['soa_ms']) else None,
            axis=1
        )
        
        if self.config['debug_mode']:
            print(f"Trial assignment complete:")
            print(f"  Inhale trials: {sum(design_df['breathphase'] == 'inhale')}")
            print(f"  Exhale trials: {sum(design_df['breathphase'] == 'exhale')}")
            print(f"  Unassigned trials: {sum(design_df['breathphase'].isna())}")
        
        return design_df
    
    def generate_participant_audio(self, participant_id, design_df, base_audio_path):
        """
        Generate participant-specific audio files.
        
        Creates two synchronized audio files:
        - Looming audio: Box breathing with embedded looming sounds
        - Tactile audio: Synchronized tactile stimuli
        
        Parameters:
        -----------
        participant_id : int
            Participant identifier
        design_df : pd.DataFrame
            Design dataframe with trial parameters
        base_audio_path : str
            Path to the base breathing audio
        
        Returns:
        --------
        tuple
            (looming_audio_path, tactile_audio_path, log_path)
        """
        # Load the base audio file
        try:
            base_audio, sample_rate = sf.read(base_audio_path)
            if self.sample_rate is None:
                self.sample_rate = sample_rate
        except Exception as e:
            print(f"Error loading base audio: {e}")
            return None, None, None
        
        # Load stimuli
        looming_stimuli, tactile_stimulus = self.load_stimuli()
        
        # Create copies of the base audio for modification
        looming_audio = base_audio.copy()
        
        # Create tactile audio with same shape as base audio
        if len(base_audio.shape) > 1:  # Stereo
            tactile_audio = np.zeros((len(base_audio), base_audio.shape[1]), dtype=base_audio.dtype)
        else:  # Mono
            tactile_audio = np.zeros(len(base_audio), dtype=base_audio.dtype)
        
        # Prepare tactile stimulus in compatible format
        if tactile_stimulus is not None:
            if len(tactile_stimulus.shape) == 1 and len(tactile_audio.shape) > 1:
                # Mono tactile, stereo audio
                tactile_stereo = np.column_stack((tactile_stimulus, tactile_stimulus))
            elif len(tactile_stimulus.shape) > 1 and len(tactile_audio.shape) == 1:
                # Stereo tactile, mono audio
                tactile_stereo = np.mean(tactile_stimulus, axis=1)
            else:
                # Same dimensions
                tactile_stereo = tactile_stimulus
        else:
            print("Warning: No tactile stimulus available")
            tactile_stereo = None
        
        # Track stimuli injections for verification
        injection_log = []
        
        # Process each trial in the design
        for _, row in design_df.iterrows():
            # Skip trials without assigned timestamps
            if pd.isnull(row['jittered_sample']):
                if self.config['debug_mode']:
                    print(f"  Skipping trial {row['trial_number']} (no timestamp)")
                continue
            
            # Get sample positions
            looming_pos = int(row['jittered_sample'])
            
            # For tactile audio, only add for regular trials (not catch)
            if row['trial_type'] != 'catch' and pd.notnull(row['soa_sample']):
                tactile_pos = int(row['soa_sample'])
            else:
                tactile_pos = None
            
            # Add looming stimulus
            stim_type = row['stimulus_type']
            if stim_type in looming_stimuli:
                looming_data = looming_stimuli[stim_type]
                
                # Check if it fits within the audio
                if looming_pos + len(looming_data) <= len(looming_audio):
                    # Handle stereo vs mono correctly
                    if len(looming_data.shape) == 2 and len(looming_audio.shape) == 2:
                        # Both stereo
                        looming_audio[looming_pos:looming_pos+len(looming_data)] += looming_data
                    elif len(looming_data.shape) == 1 and len(looming_audio.shape) == 2:
                        # Mono looming, stereo base audio
                        for ch in range(looming_audio.shape[1]):
                            looming_audio[looming_pos:looming_pos+len(looming_data), ch] += looming_data
                    elif len(looming_data.shape) == 2 and len(looming_audio.shape) == 1:
                        # Stereo looming, mono base audio
                        looming_audio[looming_pos:looming_pos+len(looming_data)] += np.mean(looming_data, axis=1)
                    else:
                        # Both mono
                        looming_audio[looming_pos:looming_pos+len(looming_data)] += looming_data
                    
                    injection_log.append({
                        'trial_number': row['trial_number'],
                        'trial_type': row['trial_type'],
                        'stimulus_type': stim_type,
                        'breathphase': row.get('breathphase', 'unknown'),
                        'looming_position': looming_pos,
                        'looming_time_ms': row['jittered_ms']
                    })
                else:
                    print(f"  Warning: Looming stimulus for trial {row['trial_number']} exceeds audio length")
            else:
                print(f"  Warning: Stimulus type '{stim_type}' not found")
            
            # Add tactile stimulus for regular trials
            if tactile_pos is not None and tactile_stereo is not None:
                # Check if it fits within the audio
                if tactile_pos + len(tactile_stereo) <= len(tactile_audio):
                    # Handle stereo vs mono correctly
                    if len(tactile_stereo.shape) == 2 and len(tactile_audio.shape) == 2:
                        # Both stereo
                        tactile_audio[tactile_pos:tactile_pos+len(tactile_stereo)] += tactile_stereo
                    elif len(tactile_stereo.shape) == 1 and len(tactile_audio.shape) == 2:
                        # Mono tactile, stereo base
                        for ch in range(tactile_audio.shape[1]):
                            tactile_audio[tactile_pos:tactile_pos+len(tactile_stereo), ch] += tactile_stereo
                    elif len(tactile_stereo.shape) == 2 and len(tactile_audio.shape) == 1:
                        # Stereo tactile, mono base
                        tactile_audio[tactile_pos:tactile_pos+len(tactile_stereo)] += np.mean(tactile_stereo, axis=1)
                    else:
                        # Both mono
                        tactile_audio[tactile_pos:tactile_pos+len(tactile_stereo)] += tactile_stereo
                    
                    # Update injection log
                    for log_entry in injection_log:
                        if log_entry['trial_number'] == row['trial_number']:
                            log_entry['tactile_position'] = tactile_pos
                            log_entry['tactile_time_ms'] = row['soa_ms']
                            log_entry['soa_ms'] = row['soa_value_ms']
                            break
                else:
                    print(f"  Warning: Tactile stimulus for trial {row['trial_number']} exceeds audio length")
        
        # Save audio files
        looming_filename = f"participant_{participant_id}_design_looming.wav"
        tactile_filename = f"participant_{participant_id}_design_tactile.wav"
        log_filename = f"participant_{participant_id}_design_stimuli_log.csv"
        
        looming_path = os.path.join(self.audio_output_dir, looming_filename)
        tactile_path = os.path.join(self.audio_output_dir, tactile_filename)
        log_path = os.path.join(self.audio_output_dir, log_filename)
        
        try:
            sf.write(looming_path, looming_audio, sample_rate)
            sf.write(tactile_path, tactile_audio, sample_rate)
            
            # Save injection log
            injection_df = pd.DataFrame(injection_log)
            injection_df.to_csv(log_path, index=False)
            
            if self.config['debug_mode']:
                print(f"Generated audio files for participant {participant_id}:")
                print(f"  Looming audio: {looming_path}")
                print(f"  Tactile audio: {tactile_path}")
                print(f"  Injection log: {log_path}")
                print(f"  Injected {len(injection_log)} stimuli")
            
            return looming_path, tactile_path, log_path
        except Exception as e:
            print(f"Error saving audio files: {e}")
            return None, None, None
    
    def verify_audio_synchronization(self, participant_id, design_df, injection_log_path):
        """
        Verify the synchronization between generated audio files and design.
        
        Parameters:
        -----------
        participant_id : int
            Participant identifier
        design_df : pd.DataFrame
            Design dataframe with trial parameters
        injection_log_path : str
            Path to the injection log file
        
        Returns:
        --------
        dict
            Verification results
        """
        results = {
            'success': False,
            'error': None,
            'warnings': [],
            'verification_details': []
        }
        
        try:
            # Load injection log
            injection_df = pd.read_csv(injection_log_path)
            
            # Count trials
            design_count = len(design_df)
            injection_count = len(injection_df)
            
            if design_count != injection_count:
                warning = f"Design has {design_count} trials but injection log has {injection_count} trials"
                results['warnings'].append(warning)
                if self.config['debug_mode']:
                    print(f"Warning: {warning}")
            
            # Verification results
            verification_details = []
            
            # Check a few sample trials for accurate timing
            sample_size = min(10, len(injection_df))
            sample_trials = injection_df.sample(n=sample_size) if len(injection_df) > 10 else injection_df
            
            for _, row in sample_trials.iterrows():
                trial_number = row['trial_number']
                
                # Find this trial in the design
                design_row = design_df[design_df['trial_number'] == trial_number]
                
                if len(design_row) == 0:
                    verification_details.append({
                        'trial_number': trial_number,
                        'status': 'error',
                        'message': f"Trial {trial_number} not found in design"
                    })
                    continue
                
                design_row = design_row.iloc[0]
                verification_item = {'trial_number': trial_number, 'status': 'ok', 'details': {}}
                
                # Check looming time
                if 'looming_time_ms' in row and 'jittered_ms' in design_row:
                    looming_diff = abs(row['looming_time_ms'] - design_row['jittered_ms'])
                    verification_item['details']['looming_diff_ms'] = looming_diff
                    
                    if looming_diff > 1:  # 1ms tolerance
                        verification_item['status'] = 'warning'
                        verification_item['details']['looming_warning'] = f"Looming timing mismatch: {looming_diff} ms"
                        results['warnings'].append(f"Trial {trial_number}: Looming timing mismatch: {looming_diff} ms")
                
                # Check tactile time for regular trials
                if design_row['trial_type'] != 'catch' and 'tactile_time_ms' in row and 'soa_ms' in design_row:
                    tactile_diff = abs(row['tactile_time_ms'] - design_row['soa_ms'])
                    verification_item['details']['tactile_diff_ms'] = tactile_diff
                    
                    if tactile_diff > 1:  # 1ms tolerance
                        verification_item['status'] = 'warning'
                        verification_item['details']['tactile_warning'] = f"Tactile timing mismatch: {tactile_diff} ms"
                        results['warnings'].append(f"Trial {trial_number}: Tactile timing mismatch: {tactile_diff} ms")
                
                # Check SOA
                if design_row['trial_type'] != 'catch' and 'soa_ms' in row and 'soa_value_ms' in design_row:
                    soa_diff = abs(row['soa_ms'] - design_row['soa_value_ms'])
                    verification_item['details']['soa_diff_ms'] = soa_diff
                    
                    if soa_diff > 1:  # 1ms tolerance
                        verification_item['status'] = 'warning'
                        verification_item['details']['soa_warning'] = f"SOA mismatch: {soa_diff} ms"
                        results['warnings'].append(f"Trial {trial_number}: SOA mismatch: {soa_diff} ms")
                
                verification_details.append(verification_item)
            
            results['verification_details'] = verification_details
            results['success'] = True
            
            if self.config['debug_mode']:
                print(f"Verification complete for participant {participant_id}")
                if results['warnings']:
                    print(f"  Warnings: {len(results['warnings'])}")
                    for warning in results['warnings'][:5]:  # Show first 5 warnings
                        print(f"    - {warning}")
                    if len(results['warnings']) > 5:
                        print(f"    - (and {len(results['warnings']) - 5} more warnings)")
                else:
                    print("  No warnings - all timings match expected values")
            
            return results
            
        except Exception as e:
            error_msg = f"Error during verification: {str(e)}"
            results['error'] = error_msg
            results['success'] = False
            if self.config['debug_mode']:
                print(error_msg)
                traceback.print_exc()
            return results
    
    def validate_counterbalancing(self, design_df):
        """
        Validate the counterbalancing of the design.
        
        Parameters:
        -----------
        design_df : pd.DataFrame
            Design dataframe with trial parameters
        
        Returns:
        --------
        dict
            Validation results
        """
        validation = {
            'counts': {},
            'balance_scores': {},
            'crosstabs': {},
            'warnings': []
        }
        
        try:
            # Count trials by type
            for factor in ['trial_type', 'stimulus_type', 'soa_value_ms', 'breathphase']:
                if factor in design_df.columns:
                    validation['counts'][factor] = design_df[factor].value_counts().to_dict()
            
            # Calculate balance scores
            def calc_balance_score(counts):
                if not counts:
                    return 0
                values = np.array(list(counts.values()))
                mean = np.mean(values)
                if mean == 0:
                    return 0
                return np.std(values) / mean
            
            for factor, counts in validation['counts'].items():
                validation['balance_scores'][factor] = calc_balance_score(counts)
            
            # Cross-tabulations
            factor_pairs = [
                ('trial_type', 'stimulus_type'),
                ('trial_type', 'breathphase'),
                ('stimulus_type', 'breathphase'),
                ('soa_value_ms', 'trial_type'),
                ('soa_value_ms', 'stimulus_type')
            ]
            
            for f1, f2 in factor_pairs:
                if f1 in design_df.columns and f2 in design_df.columns:
                    crosstab = pd.crosstab(design_df[f1], design_df[f2])
                    validation['crosstabs'][f"{f1}_vs_{f2}"] = crosstab.to_dict()
            
            # Check for major imbalances and add warnings
            for factor, balance_score in validation['balance_scores'].items():
                if balance_score > 0.2:  # 20% variation is considered imbalanced
                    warning = f"Imbalance detected in {factor}: score = {balance_score:.4f}"
                    validation['warnings'].append(warning)
                    if self.config['debug_mode']:
                        print(f"Warning: {warning}")
            
            # Additional checks for trial types vs breathing phases
            if 'trial_type' in design_df.columns and 'breathphase' in design_df.columns:
                for trial_type in design_df['trial_type'].unique():
                    type_df = design_df[design_df['trial_type'] == trial_type]
                    if 'breathphase' in type_df.columns:
                        breath_counts = type_df['breathphase'].value_counts()
                        if len(breath_counts) >= 2:
                            # Calculate max imbalance
                            max_count = breath_counts.max()
                            min_count = breath_counts.min()
                            imbalance = (max_count - min_count) / (max_count + min_count)
                            
                            if imbalance > 0.2:  # More than 20% imbalance
                                warning = f"Breath phase imbalance in {trial_type} trials: {breath_counts.to_dict()}"
                                validation['warnings'].append(warning)
                                if self.config['debug_mode']:
                                    print(f"Warning: {warning}")
            
            # Generate validation plots if enabled
            if self.config['validation']['generate_plots']:
                self.generate_validation_plots(design_df)
            
            if self.config['debug_mode']:
                print(f"Counterbalancing validation:")
                for factor, counts in validation['counts'].items():
                    print(f"  {factor}: {counts}")
                print(f"  Balance scores:")
                for factor, score in validation['balance_scores'].items():
                    status = "✓" if score <= 0.2 else "⚠"
                    print(f"    {status} {factor}: {score:.4f}")
            
            return validation
            
        except Exception as e:
            error_msg = f"Error during counterbalancing validation: {str(e)}"
            validation['error'] = error_msg
            if self.config['debug_mode']:
                print(error_msg)
                traceback.print_exc()
            return validation
    
    def generate_validation_plots(self, design_df):
        """
        Generate validation plots for the design.
        
        Parameters:
        -----------
        design_df : pd.DataFrame
            Design dataframe with trial parameters
        """
        plot_dir = self.validation_plots_dir
        
        try:
            # 1. Trial type distribution
            plt.figure(figsize=(10, 6))
            ax = sns.countplot(x='trial_type', data=design_df)
            ax.set_title('Trial Type Distribution')
            ax.set_xlabel('Trial Type')
            ax.set_ylabel('Count')
            plt.tight_layout()
            plt.savefig(os.path.join(plot_dir, 'trial_type_distribution.png'))
            plt.close()
            
            # 2. Stimulus type distribution
            plt.figure(figsize=(10, 6))
            ax = sns.countplot(x='stimulus_type', data=design_df)
            ax.set_title('Stimulus Type Distribution')
            ax.set_xlabel('Stimulus Type')
            ax.set_ylabel('Count')
            plt.tight_layout()
            plt.savefig(os.path.join(plot_dir, 'stimulus_type_distribution.png'))
            plt.close()
            
            # 3. SOA value distribution
            plt.figure(figsize=(12, 6))
            ax = sns.countplot(x='soa_value_ms', data=design_df)
            ax.set_title('SOA Value Distribution')
            ax.set_xlabel('SOA (ms)')
            ax.set_ylabel('Count')
            plt.tight_layout()
            plt.savefig(os.path.join(plot_dir, 'soa_distribution.png'))
            plt.close()
            
            # 4. Breathing phase distribution
            if 'breathphase' in design_df.columns:
                plt.figure(figsize=(8, 6))
                ax = sns.countplot(x='breathphase', data=design_df)
                ax.set_title('Breathing Phase Distribution')
                ax.set_xlabel('Breath Phase')
                ax.set_ylabel('Count')
                plt.tight_layout()
                plt.savefig(os.path.join(plot_dir, 'breathphase_distribution.png'))
                plt.close()
            
            # 5. Trial type vs breathing phase
            if 'trial_type' in design_df.columns and 'breathphase' in design_df.columns:
                plt.figure(figsize=(12, 6))
                crosstab = pd.crosstab(design_df['trial_type'], design_df['breathphase'])
                ax = crosstab.plot(kind='bar', stacked=True)
                ax.set_title('Trial Type vs. Breathing Phase')
                ax.set_xlabel('Trial Type')
                ax.set_ylabel('Count')
                plt.legend(title='Breath Phase')
                plt.tight_layout()
                plt.savefig(os.path.join(plot_dir, 'trial_type_vs_breathphase.png'))
                plt.close()
            
            # 6. Stimulus type vs SOA
            plt.figure(figsize=(14, 8))
            crosstab = pd.crosstab(design_df['stimulus_type'], design_df['soa_value_ms'])
            ax = sns.heatmap(crosstab, annot=True, cmap="YlGnBu", fmt='d')
            ax.set_title('Stimulus Type vs. SOA Value')
            plt.tight_layout()
            plt.savefig(os.path.join(plot_dir, 'stimulus_type_vs_soa.png'))
            plt.close()
            
            # 7. Trial sequence analysis
            if len(design_df) > 1:
                # Analyze trial type transitions
                trial_sequence = design_df.sort_values('trial_number')['trial_type'].tolist()
                transitions = {}
                
                for i in range(len(trial_sequence) - 1):
                    current = trial_sequence[i]
                    next_type = trial_sequence[i + 1]
                    key = (current, next_type)
                    transitions[key] = transitions.get(key, 0) + 1
                
                # Create transition matrix
                unique_types = sorted(design_df['trial_type'].unique())
                transition_matrix = np.zeros((len(unique_types), len(unique_types)))
                
                for i, type1 in enumerate(unique_types):
                    for j, type2 in enumerate(unique_types):
                        transition_matrix[i, j] = transitions.get((type1, type2), 0)
                
                # Normalize by row
                row_sums = transition_matrix.sum(axis=1, keepdims=True)
                normalized_matrix = transition_matrix / row_sums
                
                # Plot transition matrix
                plt.figure(figsize=(10, 8))
                ax = sns.heatmap(normalized_matrix, annot=True, cmap="YlGnBu", fmt='.2f',
                               xticklabels=unique_types, yticklabels=unique_types)
                ax.set_title('Trial Type Transition Matrix (Normalized)')
                ax.set_xlabel('To')
                ax.set_ylabel('From')
                plt.tight_layout()
                plt.savefig(os.path.join(plot_dir, 'trial_type_transitions.png'))
                plt.close()
                
                # 8. Trial sequence randomness test
                # Calculate runs (consecutive identical values)
                runs = 1
                for i in range(1, len(trial_sequence)):
                    if trial_sequence[i] != trial_sequence[i-1]:
                        runs += 1
                
                # Calculate expected runs for a random sequence
                trial_type_counts = pd.Series(trial_sequence).value_counts()
                n = len(trial_sequence)
                expected_runs = 1
                for count in trial_type_counts:
                    expected_runs += (count / n) * (1 - count / n) * 2 * (n - 1)
                
                # Plot runs analysis
                plt.figure(figsize=(8, 6))
                plt.bar(['Actual', 'Expected'], [runs, expected_runs], color=['blue', 'orange'])
                plt.axhline(y=expected_runs, color='r', linestyle='--')
                plt.title('Trial Sequence Randomness Analysis')
                plt.ylabel('Number of Runs')
                plt.ylim(0, max(runs, expected_runs) * 1.2)
                plt.annotate(f"Actual: {runs}", (0, runs), textcoords="offset points", 
                            xytext=(0,10), ha='center')
                plt.annotate(f"Expected: {expected_runs:.1f}", (1, expected_runs), 
                            textcoords="offset points", xytext=(0,10), ha='center')
                plt.tight_layout()
                plt.savefig(os.path.join(plot_dir, 'sequence_randomness.png'))
                plt.close()
            
            if self.config['debug_mode']:
                print(f"Generated validation plots in {plot_dir}")
                
        except Exception as e:
            print(f"Error generating validation plots: {e}")
            traceback.print_exc()
    
    def process_participant(self, participant_id):
        """
        Process a single participant from design generation to audio creation.
        
        Parameters:
        -----------
        participant_id : int
            Participant identifier
        
        Returns:
        --------
        dict
            Processing results
        """
        results = {
            'participant_id': participant_id,
            'status': 'error',
            'files': {},
            'validation': {},
            'errors': []
        }
        
        try:
            print(f"\n=== Processing participant {participant_id} ===")
            
            # Step 1: Check if audio extension is needed and extend if necessary
            if self.extended_audio_path is None:
                base_audio_path, timestamps_df = self.extend_breathing_audio()
            else:
                base_audio_path = self.extended_audio_path
                timestamps_df = self.extended_timestamps_df
            
            # Step 2: Generate counterbalanced design
            design_df = self.generate_counterbalanced_design(participant_id)
            
            # Step 3: Assign trials to breath holds
            design_df = self.assign_trials_to_breath_holds(design_df, timestamps_df)
            
            # Step 4: Save design to CSV
            design_path = os.path.join(self.experiment_log_dir, f"participant_{participant_id}_design.csv")
            design_df.to_csv(design_path, index=False)
            if self.config['debug_mode']:
                print(f"Saved design to {design_path}")
            
            results['files']['design'] = design_path
            
            # Step 5: Generate audio files
            looming_path, tactile_path, log_path = self.generate_participant_audio(
                participant_id, design_df, base_audio_path
            )
            
            if looming_path and tactile_path and log_path:
                results['files']['looming_audio'] = looming_path
                results['files']['tactile_audio'] = tactile_path
                results['files']['stimulus_log'] = log_path
            else:
                error_msg = "Failed to generate audio files"
                results['errors'].append(error_msg)
                print(f"Error: {error_msg}")
            
            # Step 6: Verify generated files
            if log_path:
                verification_results = self.verify_audio_synchronization(participant_id, design_df, log_path)
                results['validation']['synchronization'] = verification_results
                
                if verification_results.get('warnings'):
                    print(f"Warning: {len(verification_results['warnings'])} synchronization issues detected")
            
            # Step 7: Validate counterbalancing
            counterbalancing_results = self.validate_counterbalancing(design_df)
            results['validation']['counterbalancing'] = counterbalancing_results
            
            # Overall status
            if not results['errors']:
                results['status'] = 'success'
                print(f"=== Successfully processed participant {participant_id} ===\n")
            else:
                print(f"=== Completed processing for participant {participant_id} with errors ===\n")
            
            return results
            
        except Exception as e:
            error_msg = f"Error processing participant {participant_id}: {str(e)}"
            results['errors'].append(error_msg)
            print(error_msg)
            traceback.print_exc()
            return results
    
    def run(self, participant_ids=None, num_participants=None):
        """
        Run the full audio generation process for multiple participants.
        
        Parameters:
        -----------
        participant_ids : list, optional
            List of participant IDs to process
        num_participants : int, optional
            Number of participants to process (if participant_ids not provided)
        
        Returns:
        --------
        dict
            Dictionary with results for each participant
        """
        if participant_ids is None and num_participants is None:
            num_participants = 5  # Default
        
        if participant_ids is None:
            participant_ids = list(range(num_participants))
        
        results = {}
        
        print(f"=== Starting EnhancedPPSAudioGenerator ===")
        print(f"Repetitions per condition: {self.config['repetitions_per_condition']}")
        print(f"Processing {len(participant_ids)} participants")
        print(f"Output directories:")
        print(f"  Design files: {self.experiment_log_dir}")
        print(f"  Audio files: {self.audio_output_dir}")
        
        # Ensure we have calculated trial counts
        total_trials = sum(self.trial_counts.values())
        print(f"Total trials per participant: {total_trials}")
        
        # Process each participant
        for p_id in participant_ids:
            results[p_id] = self.process_participant(p_id)
        
        # Generate summary
        success_count = sum(1 for r in results.values() if r['status'] == 'success')
        print(f"\n=== Processing Summary ===")
        print(f"Successfully processed {success_count} of {len(participant_ids)} participants")
        
        # Save overall validation report
        if self.config['validation']['save_detailed_report']:
            timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
            report_path = os.path.join(self.experiment_log_dir, f"validation_report_{timestamp}.json")
            
            try:
                with open(report_path, 'w') as f:
                    # Create a simplified version of results for the report
                    report_data = {
                        'config': self.config,
                        'timestamp': timestamp,
                        'participants': {
                            p_id: {
                                'status': r['status'],
                                'validation': r['validation'],
                                'files': {k: os.path.basename(v) for k, v in r.get('files', {}).items()}
                            } for p_id, r in results.items()
                        }
                    }
                    json.dump(report_data, f, indent=2)
                
                print(f"Saved detailed validation report to {report_path}")
            except Exception as e:
                print(f"Error saving validation report: {e}")
        
        print(f"=== Completed audio generation ===")
        
        return results

# ======================================================================
# MAIN FUNCTION
# ======================================================================

def main():
    """
    Main function to run the Enhanced PPS Audio Generator.
    """
    # Create configuration
    config = {
        'repetitions_per_condition': 4,  # Change as needed
        'included_conditions': {
            'inhalation': True,
            'exhalation': True,
            'baseline': True
        },
        'catch_trial_percentage': 10,
        'debug_mode': True
    }
    
    # Create generator with custom configuration
    generator = EnhancedPPSAudioGenerator(config)
    
    # Process participants
    results = generator.run(participant_ids=[0, 1, 2, 3, 4])
    
    # Print results
    for p_id, data in results.items():
        status = data['status']
        if status == 'success':
            print(f"Participant {p_id}: Success")
            for file_type, file_path in data.get('files', {}).items():
                print(f"  {file_type}: {os.path.basename(file_path)}")
        else:
            print(f"Participant {p_id}: Error - {', '.join(data.get('errors', ['Unknown error']))}")

if __name__ == "__main__":
    main()
```

```python
import os
import numpy as np
import pandas as pd
import soundfile as sf
import re
from datetime import datetime
import random
import traceback
import json

# ======================================================================
# CONFIGURATION PARAMETERS
# ======================================================================

CONFIG = {
    'repetitions_per_condition': 4,
    'included_conditions': {
        'inhalation': True,
        'exhalation': True,
        'baseline': True,
    },
    'catch_trial_percentage': 10,
    'soa_conditions_ms': [190, 400, 700, 1000, 1500],
    'jitter_options_ms': [100, 200, 300, 400, 500],
    'debug_mode': True,
    'audio_extension': {
        'intro_segment_end': '02:00.2',
        'repeat_segment_start': '02:00.2',
        'repeat_segment_end': '04:56.4',
        'preserve_ending': True,
        'ending_duration_sec': 45,
    },
    'paths': {
        'base_dir': r"C:\Users\cogpsy-vrlab\Documents\GitHub\BreathingSpace",
        'level0_dir': r"C:\Users\cogpsy-vrlab\Documents\GitHub\BreathingSpace\Level0_StimulusFiles",
        'experiment_log_dir': r"C:\Users\cogpsy-vrlab\Documents\GitHub\BreathingSpace\ExperimentLog",
        'audio_output_dir': r"C:\Users\cogpsy-vrlab\Documents\GitHub\BreathingSpace\ExperimentAudio",
    }
}

# Updated looming stimuli paths
LOOMING_STIMULI = {
    'blue': "frontal_looming_blue_noise.wav",
    'brown': "frontal_looming_brown_noise.wav",
    'pink': "frontal_looming_pink_noise.wav",
    'white': "frontal_looming_white_noise.wav"
}

# ======================================================================
# PPS AUDIO GENERATOR
# ======================================================================

class PPSAudioGenerator:
    """Generates audio files for Peripersonal Space (PPS) experiments."""
    
    def __init__(self, config=None):
        """Initialize with configuration parameters."""
        self.config = config or CONFIG
        self.setup_paths()
        self.trial_counts = self._calculate_trial_counts()
        self.conditions = [cond for cond, include in self.config['included_conditions'].items() if include]
        
        # Initialize properties
        self.timestamps_df = None
        self.extended_audio = None
        self.extended_audio_path = None
        self.extended_timestamps_df = None
        self.sample_rate = None
        self.looming_stimuli = {}
        self.tactile_stimulus = None
        
        if self.config['debug_mode']:
            print(f"Initialized PPSAudioGenerator with:")
            print(f"  - Repetitions: {self.config['repetitions_per_condition']}")
            print(f"  - Conditions: {self.conditions}")
            print(f"  - SOA values: {self.config['soa_conditions_ms']}")
            print(f"  - Total trials per participant: {sum(self.trial_counts.values())}")
    
    def setup_paths(self):
        """Set up file paths for input and output files."""
        # Get base paths from config
        base_dir = self.config['paths']['base_dir']
        level0_dir = self.config['paths']['level0_dir']
        self.experiment_log_dir = self.config['paths']['experiment_log_dir']
        self.audio_output_dir = self.config['paths']['audio_output_dir']
        
        # Input files
        self.breathing_audio_path = os.path.join(level0_dir, "BoxBreathing.wav")
        self.timestamps_csv_path = os.path.join(level0_dir, "BoxBreathing_timestamps.csv")
        
        # Stimulus files
        self.looming_stimuli_paths = {}
        for stim_type, filename in LOOMING_STIMULI.items():
            self.looming_stimuli_paths[stim_type] = os.path.join(level0_dir, filename)
        
        self.tactile_stimulus_path = os.path.join(level0_dir, "tactile_stimulus.wav")
        
        # Create output directories
        os.makedirs(self.experiment_log_dir, exist_ok=True)
        os.makedirs(self.audio_output_dir, exist_ok=True)
        
        if self.config['debug_mode']:
            print(f"Set up paths:")
            print(f"  - Breathing audio: {self.breathing_audio_path}")
            print(f"  - Timestamps CSV: {self.timestamps_csv_path}")
            print(f"  - Looming stimuli: {len(self.looming_stimuli_paths)} files")
            print(f"  - Tactile stimulus: {self.tactile_stimulus_path}")
            print(f"  - Output directories created")
    
    def _calculate_trial_counts(self):
        """Calculate trial counts based on the configuration."""
        # Get parameters
        reps = self.config['repetitions_per_condition']
        soa_count = len(self.config['soa_conditions_ms'])
        stimulus_types = list(LOOMING_STIMULI.keys())
        catch_percentage = self.config['catch_trial_percentage']
        
        # Calculate trials for each condition
        trial_counts = {}
        for condition, include in self.config['included_conditions'].items():
            if include:
                # For each condition, we need (# stimulus types) × (# SOA values) × (repetitions) trials
                trial_counts[condition] = len(stimulus_types) * soa_count * reps
        
        # Calculate catch trials
        regular_trials = sum(trial_counts.values())
        catch_trials = int(np.ceil(regular_trials * catch_percentage / 100))
        trial_counts['catch'] = catch_trials
        
        if self.config['debug_mode']:
            print(f"Trial count calculation:")
            for cond, count in trial_counts.items():
                print(f"  {cond}: {count} trials")
            print(f"  Total: {sum(trial_counts.values())} trials")
        
        return trial_counts
    
    def parse_timestamps(self):
        """Parse the box breathing timestamps CSV."""
        if self.timestamps_df is not None:
            return self.timestamps_df
        
        try:
            # Read the CSV file
            df = pd.read_csv(self.timestamps_csv_path)
            
            # Add breathphase if not present
            if 'breathphase' not in df.columns:
                # Infer from pattern (even indices are inhale, odd are exhale)
                df['breathphase'] = ['inhale' if i % 2 == 0 else 'exhale' for i in range(len(df))]
            
            # Add time_seconds column (milliseconds converted to seconds)
            df['time_seconds'] = df['milliseconds'] / 1000
            
            # Store the processed dataframe
            self.timestamps_df = df
            
            if self.config['debug_mode']:
                inhale_count = sum(df['breathphase'] == 'inhale')
                exhale_count = sum(df['breathphase'] == 'exhale')
                print(f"Parsed timestamps: {len(df)} total entries")
                print(f"Breath phases: {inhale_count} inhale, {exhale_count} exhale")
            
            return df
            
        except Exception as e:
            print(f"Error parsing timestamps: {e}")
            traceback.print_exc()
            return pd.DataFrame()
    
    def analyze_audio_extension_needs(self):
        """Analyze whether audio extension is needed based on trial count."""
        # Parse timestamps
        timestamps_df = self.parse_timestamps()
        
        # Calculate required trials
        total_trials = sum(self.trial_counts.values())
        
        # We need breath holds for all trials except catch trials
        required_breath_holds = total_trials - self.trial_counts.get('catch', 0)
        
        # Get segmentation indices
        intro_end = self.config['audio_extension']['intro_segment_end']
        repeat_start = self.config['audio_extension']['repeat_segment_start']
        repeat_end = self.config['audio_extension']['repeat_segment_end']
        
        # Find indicies in the dataframe
        intro_match = timestamps_df[timestamps_df['timestamp'] == intro_end]
        repeat_start_match = timestamps_df[timestamps_df['timestamp'] == repeat_start]
        repeat_end_match = timestamps_df[timestamps_df['timestamp'] == repeat_end]
        
        intro_end_idx = intro_match.index[0] if not intro_match.empty else None
        repeat_start_idx = repeat_start_match.index[0] if not repeat_start_match.empty else None
        repeat_end_idx = repeat_end_match.index[0] if not repeat_end_match.empty else None
        
        if intro_end_idx is None or repeat_start_idx is None or repeat_end_idx is None:
            # Use default values (reserve first 10% of timestamps)
            total_timestamps = len(timestamps_df)
            intro_end_idx = int(total_timestamps * 0.1)
            repeat_start_idx = intro_end_idx
            repeat_end_idx = int(total_timestamps * 0.8)
        
        # Calculate available usable breath holds (after intro, before ending)
        available_breath_holds = len(timestamps_df) - (intro_end_idx + 1)
        
        # Check if extension is needed
        extension_needed = required_breath_holds > available_breath_holds
        
        if self.config['debug_mode']:
            print(f"Audio extension analysis:")
            print(f"  Required breath holds: {required_breath_holds}")
            print(f"  Available breath holds: {available_breath_holds}")
            print(f"  Extension needed: {extension_needed}")
            print(f"  Segmentation indices: intro_end={intro_end_idx}, "
                  f"repeat_start={repeat_start_idx}, repeat_end={repeat_end_idx}")
        
        return (
            extension_needed, 
            required_breath_holds, 
            available_breath_holds, 
            (intro_end_idx, repeat_start_idx, repeat_end_idx)
        )
    
    def extend_breathing_audio(self):
        """Extend the box breathing audio to accommodate all trials."""
        # Check if extension is needed
        extension_info = self.analyze_audio_extension_needs()
        extension_needed, required_holds, available_holds, segment_indices = extension_info
        intro_end_idx, repeat_start_idx, repeat_end_idx = segment_indices
        
        # Parse timestamps if not already done
        timestamps_df = self.parse_timestamps()
        
        if not extension_needed:
            if self.config['debug_mode']:
                print("Audio extension not needed, using original audio file.")
            return self.breathing_audio_path, timestamps_df
        
        # Load original audio
        original_audio, sample_rate = sf.read(self.breathing_audio_path)
        self.sample_rate = sample_rate
        
        # Get sample indices for these boundaries
        intro_end_sample = int(timestamps_df.iloc[intro_end_idx]['sample_index'])
        repeat_start_sample = int(timestamps_df.iloc[repeat_start_idx]['sample_index'])
        repeat_end_sample = int(timestamps_df.iloc[repeat_end_idx]['sample_index'])
        
        # Calculate segment size in breath holds
        segment_size = repeat_end_idx - repeat_start_idx + 1
        segment_samples = repeat_end_sample - repeat_start_sample
        
        # Calculate how many additional segments we need
        additional_holds_needed = required_holds - available_holds
        additional_segments_needed = int(np.ceil(additional_holds_needed / segment_size))
        
        if self.config['debug_mode']:
            print(f"Audio extension details:")
            print(f"  Segment size: {segment_size} breath holds ({segment_samples} samples)")
            print(f"  Additional holds needed: {additional_holds_needed}")
            print(f"  Additional segments needed: {additional_segments_needed}")
        
        # Create extended audio
        # Keep everything up to the intro_end
        extended_audio = original_audio[:intro_end_sample].copy()
        
        # Add the number of segments needed
        for i in range(additional_segments_needed + 1):  # +1 to keep the original segment
            segment = original_audio[repeat_start_sample:repeat_end_sample].copy()
            if len(extended_audio.shape) > 1 and len(segment.shape) > 1:
                # Both are stereo
                extended_audio = np.vstack([extended_audio, segment])
            else:
                # At least one is mono
                if len(extended_audio.shape) > 1:
                    # Extended is stereo, segment is mono
                    segment_stereo = np.column_stack([segment, segment])
                    extended_audio = np.vstack([extended_audio, segment_stereo])
                elif len(segment.shape) > 1:
                    # Extended is mono, segment is stereo
                    extended_mono = np.column_stack([extended_audio, extended_audio])
                    segment_mono = np.mean(segment, axis=1)
                    extended_audio = np.vstack([extended_mono, segment_mono])
                else:
                    # Both are mono
                    extended_audio = np.concatenate([extended_audio, segment])
        
        # Add the remaining part if we want to preserve the ending
        if self.config['audio_extension']['preserve_ending']:
            ending_duration_sec = self.config['audio_extension']['ending_duration_sec']
            ending_samples = min(
                len(original_audio) - repeat_end_sample,
                int(ending_duration_sec * sample_rate)
            )
            
            ending_segment = original_audio[repeat_end_sample:repeat_end_sample+ending_samples].copy()
            
            # Add ending segment
            if len(extended_audio.shape) > 1 and len(ending_segment.shape) > 1:
                # Both are stereo
                extended_audio = np.vstack([extended_audio, ending_segment])
            else:
                # At least one is mono
                if len(extended_audio.shape) > 1:
                    # Extended is stereo, ending is mono
                    ending_stereo = np.column_stack([ending_segment, ending_segment])
                    extended_audio = np.vstack([extended_audio, ending_stereo])
                elif len(ending_segment.shape) > 1:
                    # Extended is mono, ending is stereo
                    extended_mono = np.column_stack([extended_audio, extended_audio])
                    ending_mono = np.mean(ending_segment, axis=1)
                    extended_audio = np.vstack([extended_mono, ending_mono])
                else:
                    # Both are mono
                    extended_audio = np.concatenate([extended_audio, ending_segment])
        
        # Create extended timestamps
        # Start with original timestamps up to intro_end
        extended_timestamps = timestamps_df.iloc[:intro_end_idx+1].copy()
        
        # Add the extended segments
        time_offset = 0
        sample_offset = 0
        milliseconds_offset = 0
        
        for i in range(additional_segments_needed + 1):  # +1 to keep the original segment
            segment_timestamps = timestamps_df.iloc[repeat_start_idx:repeat_end_idx+1].copy()
            
            if i > 0:  # Skip offset for first segment (it's the original one)
                # Calculate offsets
                segment_duration_sec = (timestamps_df.iloc[repeat_end_idx]['time_seconds'] - 
                                    timestamps_df.iloc[repeat_start_idx]['time_seconds'])
                segment_milliseconds = (timestamps_df.iloc[repeat_end_idx]['milliseconds'] - 
                                      timestamps_df.iloc[repeat_start_idx]['milliseconds'])
                
                time_offset += segment_duration_sec
                milliseconds_offset += segment_milliseconds
                sample_offset += segment_samples
                
                # Apply offsets to this segment
                segment_timestamps['time_seconds'] += time_offset
                segment_timestamps['milliseconds'] += milliseconds_offset
                segment_timestamps['sample_index'] += sample_offset
                
                # Update timestamp strings
                for idx, row in segment_timestamps.iterrows():
                    minutes = int(row['time_seconds'] // 60)
                    seconds = row['time_seconds'] % 60
                    segment_timestamps.at[idx, 'timestamp'] = f"{minutes:02d}:{seconds:.1f}"
            
            # Append this segment to extended timestamps
            extended_timestamps = pd.concat([extended_timestamps, segment_timestamps])
        
        # Add ending timestamps if preserving ending
        if self.config['audio_extension']['preserve_ending']:
            remaining_timestamps = timestamps_df.iloc[repeat_end_idx+1:].copy()
            
            if not remaining_timestamps.empty:
                # Apply final offsets to remaining
                remaining_timestamps['time_seconds'] += time_offset
                remaining_timestamps['milliseconds'] += milliseconds_offset
                remaining_timestamps['sample_index'] += sample_offset
                
                # Update timestamp strings for remaining
                for idx, row in remaining_timestamps.iterrows():
                    minutes = int(row['time_seconds'] // 60)
                    seconds = row['time_seconds'] % 60
                    remaining_timestamps.at[idx, 'timestamp'] = f"{minutes:02d}:{seconds:.1f}"
                
                # Append remaining
                extended_timestamps = pd.concat([extended_timestamps, remaining_timestamps])
        
        # Reset index
        extended_timestamps = extended_timestamps.reset_index(drop=True)
        
        # Save extended audio file
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        extended_audio_path = os.path.join(
            self.audio_output_dir, 
            f"extended_boxbreathing_{timestamp}.wav"
        )
        sf.write(extended_audio_path, extended_audio, sample_rate)
        
        # Save extended timestamps
        extended_timestamps_path = os.path.join(
            self.experiment_log_dir, 
            f"extended_boxbreathing_timestamps_{timestamp}.csv"
        )
        extended_timestamps.to_csv(extended_timestamps_path, index=False)
        
        if self.config['debug_mode']:
            print(f"Extended audio created:")
            print(f"  Original length: {len(original_audio)/sample_rate:.2f} seconds")
            print(f"  Extended length: {len(extended_audio)/sample_rate:.2f} seconds")
            print(f"  Extended audio saved to: {extended_audio_path}")
            print(f"  Extended timestamps saved to: {extended_timestamps_path}")
        
        # Store for later use
        self.extended_audio = extended_audio
        self.extended_audio_path = extended_audio_path
        self.extended_timestamps_df = extended_timestamps
        
        return extended_audio_path, extended_timestamps
    
    def load_stimuli(self):
        """Load all stimulus files."""
        if self.looming_stimuli and self.tactile_stimulus is not None:
            return self.looming_stimuli, self.tactile_stimulus
        
        # Load looming stimuli
        looming_stimuli = {}
        for stim_type, path in self.looming_stimuli_paths.items():
            try:
                data, sr = sf.read(path)
                if self.sample_rate is None:
                    self.sample_rate = sr
                elif sr != self.sample_rate:
                    print(f"Warning: Sample rate mismatch for {stim_type}: {sr} vs {self.sample_rate}")
                looming_stimuli[stim_type] = data
                if self.config['debug_mode']:
                    print(f"Loaded {stim_type} looming stimulus: {len(data)} samples")
            except Exception as e:
                print(f"Error loading {stim_type} stimulus: {e}")
        
        # Load tactile stimulus
        try:
            tactile_data, sr = sf.read(self.tactile_stimulus_path)
            if self.sample_rate is None:
                self.sample_rate = sr
            elif sr != self.sample_rate:
                print(f"Warning: Sample rate mismatch for tactile: {sr} vs {self.sample_rate}")
            self.tactile_stimulus = tactile_data
            if self.config['debug_mode']:
                print(f"Loaded tactile stimulus: {len(tactile_data)} samples")
        except Exception as e:
            print(f"Error loading tactile stimulus: {e}")
            self.tactile_stimulus = None
        
        # Store for later use
        self.looming_stimuli = looming_stimuli
        
        return looming_stimuli, self.tactile_stimulus
    
    def generate_counterbalanced_design(self, participant_id):
        """Generate a counterbalanced design for a participant."""
        # Get parameters
        soa_conditions = self.config['soa_conditions_ms']
        jitter_options = self.config['jitter_options_ms']
        stimulus_types = list(LOOMING_STIMULI.keys())
        
        # Create lists to hold all trials
        all_trials = []
        
        # Generate regular trials for each condition
        for condition, count in self.trial_counts.items():
            # Skip catch trials (handled separately) or conditions with zero trials
            if condition == 'catch' or count == 0:
                continue
            
            # Create balanced combinations of stimulus type and SOA
            stim_soa_combinations = list(product(stimulus_types, soa_conditions))
            
            # Ensure proper repetition
            reps_needed = count // len(stim_soa_combinations)
            remainder = count % len(stim_soa_combinations)
            
            # Use full repetitions
            full_combinations = stim_soa_combinations * reps_needed
            
            # Add remainder trials (randomly selected to maintain balance)
            remainder_combinations = random.sample(stim_soa_combinations, remainder)
            
            # Combine all combinations
            all_combinations = full_combinations + remainder_combinations
            
            # Shuffle to randomize order
            random.shuffle(all_combinations)
            
            # Create trial entries
            for i, (stim_type, soa) in enumerate(all_combinations):
                all_trials.append({
                    'participant_id': participant_id,
                    'trial_number': len(all_trials) + 1,
                    'stimulus_type': stim_type,
                    'trial_type': condition,
                    'soa_condition_idx': soa_conditions.index(soa),
                    'soa_value_ms': soa,
                    'jitter_ms': random.choice(jitter_options),
                    'is_tactile': True  # Regular trials have tactile component
                })
        
        # Add catch trials
        catch_count = self.trial_counts.get('catch', 0)
        if catch_count > 0:
            # Create balanced catch trials (no tactile stimulus)
            catch_stim_combinations = list(product(stimulus_types, soa_conditions))
            
            # Ensure enough combinations
            while len(catch_stim_combinations) < catch_count:
                catch_stim_combinations.extend(catch_stim_combinations)
            
            # Select random subset if too many
            catch_combinations = random.sample(catch_stim_combinations, catch_count)
            
            # Create catch trial entries
            for i, (stim_type, soa) in enumerate(catch_combinations):
                all_trials.append({
                    'participant_id': participant_id,
                    'trial_number': len(all_trials) + 1,
                    'stimulus_type': stim_type,
                    'trial_type': 'catch',
                    'soa_condition_idx': soa_conditions.index(soa),
                    'soa_value_ms': soa,
                    'jitter_ms': random.choice(jitter_options),
                    'is_tactile': False  # Catch trials have no tactile component
                })
        
        # Create DataFrame
        design_df = pd.DataFrame(all_trials)
        
        # Shuffle the design to randomize trial order
        # But use a consistent seed based on participant ID for reproducibility
        design_df = design_df.sample(frac=1, random_state=participant_id*100).reset_index(drop=True)
        
        # Update trial numbers after shuffling
        design_df['trial_number'] = range(1, len(design_df) + 1)
        
        return design_df
    
    def assign_trials_to_breath_holds(self, design_df, timestamps_df):
        """Assign trials to specific breath holds while maintaining balance."""
        # Get audio extension parameters
        intro_end = self.config['audio_extension']['intro_segment_end']
        
        # Find intro_end_idx in timestamps
        intro_end_match = timestamps_df[timestamps_df['timestamp'] == intro_end]
        intro_end_idx = intro_end_match.index[0] if not intro_end_match.empty else None
        
        if intro_end_idx is None:
            # Fallback to a percentage of total timestamps
            intro_end_idx = int(len(timestamps_df) * 0.1)
            print(f"Warning: Using fallback intro_end_idx = {intro_end_idx}")
        
        # Separate inhale and exhale timestamps
        # Skip first few timestamps (introduction period)
        usable_timestamps = timestamps_df.iloc[intro_end_idx+1:].copy()
        inhale_timestamps = usable_timestamps[usable_timestamps['breathphase'] == 'inhale']
        exhale_timestamps = usable_timestamps[usable_timestamps['breathphase'] == 'exhale']
        
        if self.config['debug_mode']:
            print(f"Assigning trials to breath holds:")
            print(f"  Available inhale holds: {len(inhale_timestamps)}")
            print(f"  Available exhale holds: {len(exhale_timestamps)}")
        
        # Create copies for assignment
        inhale_indices = inhale_timestamps.index.tolist()
        exhale_indices = exhale_timestamps.index.tolist()
        
        # Random shuffle both lists
        random.shuffle(inhale_indices)
        random.shuffle(exhale_indices)
        
        # Assign trials to breath holds
        trial_assignments = {}
        
        # Group trials by condition
        for condition in set(design_df['trial_type']):
            condition_trials = design_df[design_df['trial_type'] == condition]
            condition_count = len(condition_trials)
            
            # Calculate how many trials of each breath phase we need
            inhale_count = condition_count // 2
            exhale_count = condition_count - inhale_count
            
            if self.config['debug_mode']:
                print(f"  Condition '{condition}': {condition_count} trials - {inhale_count} inhale, {exhale_count} exhale")
            
            if inhale_count > len(inhale_indices) or exhale_count > len(exhale_indices):
                print(f"Warning: Not enough breath holds for condition '{condition}'")
                print(f"  Need {inhale_count} inhale holds, have {len(inhale_indices)}")
                print(f"  Need {exhale_count} exhale holds, have {len(exhale_indices)}")
                
                # Adjust to use what we have
                inhale_count = min(inhale_count, len(inhale_indices))
                exhale_count = min(exhale_count, len(exhale_indices))
            
            # Assign to inhale holds
            for idx, trial_idx in zip(inhale_indices[:inhale_count], condition_trials.index[:inhale_count]):
                timestamp_row = timestamps_df.loc[idx]
                trial_assignments[trial_idx] = {
                    'timestamp_idx': idx,
                    'breathphase': 'inhale',
                    'timestamp': timestamp_row['timestamp'],
                    'milliseconds': timestamp_row['milliseconds'],
                    'sample_index': timestamp_row['sample_index'],
                    'time_seconds': timestamp_row['time_seconds']
                }
            
            # Assign to exhale holds
            for idx, trial_idx in zip(exhale_indices[:exhale_count], condition_trials.index[inhale_count:inhale_count+exhale_count]):
                timestamp_row = timestamps_df.loc[idx]
                trial_assignments[trial_idx] = {
                    'timestamp_idx': idx,
                    'breathphase': 'exhale',
                    'timestamp': timestamp_row['timestamp'],
                    'milliseconds': timestamp_row['milliseconds'],
                    'sample_index': timestamp_row['sample_index'],
                    'time_seconds': timestamp_row['time_seconds'] 
                }
            
            # Remove assigned indices to prevent reuse
            inhale_indices = inhale_indices[inhale_count:]
            exhale_indices = exhale_indices[exhale_count:]
        
        # Add assignment info to design dataframe
        design_df['breathphase'] = design_df.index.map(lambda idx: trial_assignments.get(idx, {}).get('breathphase'))
        design_df['timestamp'] = design_df.index.map(lambda idx: trial_assignments.get(idx, {}).get('timestamp'))
        design_df['milliseconds'] = design_df.index.map(lambda idx: trial_assignments.get(idx, {}).get('milliseconds'))
        design_df['sample_index'] = design_df.index.map(lambda idx: trial_assignments.get(idx, {}).get('sample_index'))
        design_df['time_seconds'] = design_df.index.map(lambda idx: trial_assignments.get(idx, {}).get('time_seconds'))
        
        # Handle any unassigned trials with fallback assignment
        unassigned = design_df[design_df['breathphase'].isna()]
        if not unassigned.empty:
            print(f"Warning: {len(unassigned)} trials could not be assigned to breath holds")
            
            # Fallback assignment - assign to random timestamps
            available_timestamps = timestamps_df.loc[
                ~timestamps_df.index.isin([a['timestamp_idx'] for a in trial_assignments.values() if 'timestamp_idx' in a])
            ]
            
            if not available_timestamps.empty:
                print(f"  Using fallback assignment with {len(available_timestamps)} available timestamps")
                for idx in unassigned.index:
                    if available_timestamps.empty:
                        break
                    
                 
                    # Get a random timestamp
                    ts_idx = random.choice(available_timestamps.index)
                    timestamp_row = available_timestamps.loc[ts_idx]
                    
                    # Assign to this trial
                    design_df.at[idx, 'breathphase'] = timestamp_row.get('breathphase', 'unknown')
                    design_df.at[idx, 'timestamp'] = timestamp_row['timestamp']
                    design_df.at[idx, 'milliseconds'] = timestamp_row['milliseconds']
                    design_df.at[idx, 'sample_index'] = timestamp_row['sample_index']
                    design_df.at[idx, 'time_seconds'] = timestamp_row['time_seconds']
                    
                    # Remove this timestamp from available ones
                    available_timestamps = available_timestamps.drop(ts_idx)
        
        # Calculate timestamps with jitter
        design_df['jittered_ms'] = design_df.apply(
            lambda row: row['milliseconds'] + row['jitter_ms'] if pd.notnull(row['milliseconds']) else None,
            axis=1
        )
        
        # Calculate timestamps with SOA for tactile stimulus (only for non-catch trials)
        design_df['soa_ms'] = design_df.apply(
            lambda row: row['jittered_ms'] + row['soa_value_ms'] if pd.notnull(row['jittered_ms']) and row['trial_type'] != 'catch' else None, 
            axis=1
        )
        
        # Format timestamps for output
        design_df['timestamp_original'] = design_df.apply(
            lambda row: f"{int(row['milliseconds'] // 60000):02}:{(row['milliseconds'] % 60000) / 1000:.1f}" if pd.notnull(row['milliseconds']) else None,
            axis=1
        )
        
        design_df['timestamp_after_jitter'] = design_df.apply(
            lambda row: f"{int(row['jittered_ms'] // 60000):02}:{(row['jittered_ms'] % 60000) / 1000:.1f}" if pd.notnull(row['jittered_ms']) else None,
            axis=1
        )
        
        design_df['timestamp_with_soa'] = design_df.apply(
            lambda row: f"{int(row['soa_ms'] // 60000):02}:{(row['soa_ms'] % 60000) / 1000:.1f}" if pd.notnull(row['soa_ms']) else None,
            axis=1
        )
        
        # Convert milliseconds to sample indices
        sample_rate = self.sample_rate if self.sample_rate else 48000
        design_df['base_ms'] = design_df['milliseconds']
        design_df['jittered_sample'] = design_df.apply(
            lambda row: int(row['jittered_ms'] / 1000 * sample_rate) if pd.notnull(row['jittered_ms']) else None,
            axis=1
        )
        design_df['soa_sample'] = design_df.apply(
            lambda row: int(row['soa_ms'] / 1000 * sample_rate) if pd.notnull(row['soa_ms']) else None,
            axis=1
        )
        
        return design_df
    
    def generate_participant_audio(self, participant_id, design_df, base_audio_path):
        """Generate participant-specific audio files."""
        # Load the base audio file
        try:
            base_audio, sample_rate = sf.read(base_audio_path)
            if self.sample_rate is None:
                self.sample_rate = sample_rate
        except Exception as e:
            print(f"Error loading base audio: {e}")
            return None, None, None
        
        # Load stimuli
        looming_stimuli, tactile_stimulus = self.load_stimuli()
        
        # Create copies of the base audio for modification
        looming_audio = base_audio.copy()
        
        # Create tactile audio with same shape as base audio
        if len(base_audio.shape) > 1:  # Stereo
            tactile_audio = np.zeros((len(base_audio), base_audio.shape[1]), dtype=base_audio.dtype)
        else:  # Mono
            tactile_audio = np.zeros(len(base_audio), dtype=base_audio.dtype)
        
        # Prepare tactile stimulus in compatible format
        if tactile_stimulus is not None:
            if len(tactile_stimulus.shape) == 1 and len(tactile_audio.shape) > 1:
                # Mono tactile, stereo audio
                tactile_stereo = np.column_stack((tactile_stimulus, tactile_stimulus))
            elif len(tactile_stimulus.shape) > 1 and len(tactile_audio.shape) == 1:
                # Stereo tactile, mono audio
                tactile_stereo = np.mean(tactile_stimulus, axis=1)
            else:
                # Same dimensions
                tactile_stereo = tactile_stimulus
        else:
            print("Warning: No tactile stimulus available")
            tactile_stereo = None
        
        # Track stimuli injections for verification
        injection_log = []
        
        # Process each trial in the design
        for _, row in design_df.iterrows():
            # Skip trials without assigned timestamps
            if pd.isnull(row['jittered_sample']):
                continue
            
            # Get sample positions
            looming_pos = int(row['jittered_sample'])
            
            # For tactile audio, only add for regular trials (not catch)
            if row['trial_type'] != 'catch' and pd.notnull(row['soa_sample']):
                tactile_pos = int(row['soa_sample'])
            else:
                tactile_pos = None
            
            # Add looming stimulus
            stim_type = row['stimulus_type']
            if stim_type in looming_stimuli:
                looming_data = looming_stimuli[stim_type]
                
                # Check if it fits within the audio
                if looming_pos + len(looming_data) <= len(looming_audio):
                    # Handle stereo vs mono correctly
                    if len(looming_data.shape) == 2 and len(looming_audio.shape) == 2:
                        # Both stereo
                        looming_audio[looming_pos:looming_pos+len(looming_data)] += looming_data
                    elif len(looming_data.shape) == 1 and len(looming_audio.shape) == 2:
                        # Mono looming, stereo base audio
                        for ch in range(looming_audio.shape[1]):
                            looming_audio[looming_pos:looming_pos+len(looming_data), ch] += looming_data
                    elif len(looming_data.shape) == 2 and len(looming_audio.shape) == 1:
                        # Stereo looming, mono base audio
                        looming_audio[looming_pos:looming_pos+len(looming_data)] += np.mean(looming_data, axis=1)
                    else:
                        # Both mono
                        looming_audio[looming_pos:looming_pos+len(looming_data)] += looming_data
                    
                    injection_log.append({
                        'trial_number': row['trial_number'],
                        'trial_type': row['trial_type'],
                        'stimulus_type': stim_type,
                        'breathphase': row.get('breathphase', 'unknown'),
                        'looming_position': looming_pos,
                        'looming_time_ms': row['jittered_ms']
                    })
            
            # Add tactile stimulus for regular trials
            if tactile_pos is not None and tactile_stereo is not None:
                # Check if it fits within the audio
                if tactile_pos + len(tactile_stereo) <= len(tactile_audio):
                    # Handle stereo vs mono correctly
                    if len(tactile_stereo.shape) == 2 and len(tactile_audio.shape) == 2:
                        # Both stereo
                        tactile_audio[tactile_pos:tactile_pos+len(tactile_stereo)] += tactile_stereo
                    elif len(tactile_stereo.shape) == 1 and len(tactile_audio.shape) == 2:
                        # Mono tactile, stereo base
                        for ch in range(tactile_audio.shape[1]):
                            tactile_audio[tactile_pos:tactile_pos+len(tactile_stereo), ch] += tactile_stereo
                    elif len(tactile_stereo.shape) == 2 and len(tactile_audio.shape) == 1:
                        # Stereo tactile, mono base
                        tactile_audio[tactile_pos:tactile_pos+len(tactile_stereo)] += np.mean(tactile_stereo, axis=1)
                    else:
                        # Both mono
                        tactile_audio[tactile_pos:tactile_pos+len(tactile_stereo)] += tactile_stereo
                    
                    # Update injection log
                    for log_entry in injection_log:
                        if log_entry['trial_number'] == row['trial_number']:
                            log_entry['tactile_position'] = tactile_pos
                            log_entry['tactile_time_ms'] = row['soa_ms']
                            log_entry['soa_ms'] = row['soa_value_ms']
                            break
        
        # Save audio files
        looming_filename = f"participant_{participant_id}_design_looming.wav"
        tactile_filename = f"participant_{participant_id}_design_tactile.wav"
        log_filename = f"participant_{participant_id}_design_stimuli_log.csv"
        
        looming_path = os.path.join(self.audio_output_dir, looming_filename)
        tactile_path = os.path.join(self.audio_output_dir, tactile_filename)
        log_path = os.path.join(self.audio_output_dir, log_filename)
        
        try:
            sf.write(looming_path, looming_audio, sample_rate)
            sf.write(tactile_path, tactile_audio, sample_rate)
            
            # Save injection log
            injection_df = pd.DataFrame(injection_log)
            injection_df.to_csv(log_path, index=False)
            
            if self.config['debug_mode']:
                print(f"Generated audio files for participant {participant_id}:")
                print(f"  Looming audio: {looming_path}")
                print(f"  Tactile audio: {tactile_path}")
                print(f"  Injection log: {log_path}")
                print(f"  Injected {len(injection_log)} stimuli")
            
            return looming_path, tactile_path, log_path
        except Exception as e:
            print(f"Error saving audio files: {e}")
            return None, None, None
    
    def verify_audio_synchronization(self, participant_id, design_df, injection_log_path):
        """Verify the synchronization between generated audio files and design."""
        results = {
            'success': False,
            'error': None,
            'warnings': []
        }
        
        try:
            # Load injection log
            injection_df = pd.read_csv(injection_log_path)
            
            # Count trials
            design_count = len(design_df)
            injection_count = len(injection_df)
            
            if design_count != injection_count:
                warning = f"Design has {design_count} trials but injection log has {injection_count} trials"
                results['warnings'].append(warning)
            
            # Check a few sample trials for accurate timing
            sample_size = min(5, len(injection_df))
            sample_trials = injection_df.sample(n=sample_size) if len(injection_df) > 5 else injection_df
            
            for _, row in sample_trials.iterrows():
                trial_number = row['trial_number']
                
                # Find this trial in the design
                design_row = design_df[design_df['trial_number'] == trial_number]
                
                if len(design_row) == 0:
                    results['warnings'].append(f"Trial {trial_number} not found in design")
                    continue
                
                design_row = design_row.iloc[0]
                
                # Check looming time
                if 'looming_time_ms' in row and 'jittered_ms' in design_row:
                    looming_diff = abs(row['looming_time_ms'] - design_row['jittered_ms'])
                    
                    if looming_diff > 1:  # 1ms tolerance
                        results['warnings'].append(f"Trial {trial_number}: Looming timing mismatch: {looming_diff} ms")
                
                # Check tactile time for regular trials
                if design_row['trial_type'] != 'catch' and 'tactile_time_ms' in row and 'soa_ms' in design_row:
                    tactile_diff = abs(row['tactile_time_ms'] - design_row['soa_ms'])
                    
                    if tactile_diff > 1:  # 1ms tolerance
                        results['warnings'].append(f"Trial {trial_number}: Tactile timing mismatch: {tactile_diff} ms")
            
            results['success'] = True
            
            if self.config['debug_mode']:
                print(f"Verification complete for participant {participant_id}")
                if results['warnings']:
                    print(f"  Warnings: {len(results['warnings'])}")
                    for warning in results['warnings'][:3]:  # Show first 3 warnings
                        print(f"    - {warning}")
                    if len(results['warnings']) > 3:
                        print(f"    - (and {len(results['warnings']) - 3} more warnings)")
                else:
                    print("  No warnings - all timings match expected values")
            
            return results
            
        except Exception as e:
            results['error'] = f"Error during verification: {str(e)}"
            results['success'] = False
            if self.config['debug_mode']:
                print(results['error'])
            return results
    
    def process_participant(self, participant_id):
        """Process a single participant from design generation to audio creation."""
        results = {
            'participant_id': participant_id,
            'status': 'error',
            'files': {},
            'errors': []
        }
        
        try:
            print(f"\n=== Processing participant {participant_id} ===")
            
            # Step 1: Check if audio extension is needed and extend if necessary
            if self.extended_audio_path is None:
                base_audio_path, timestamps_df = self.extend_breathing_audio()
            else:
                base_audio_path = self.extended_audio_path
                timestamps_df = self.extended_timestamps_df
            
            # Step 2: Generate counterbalanced design
            design_df = self.generate_counterbalanced_design(participant_id)
            
            # Step 3: Assign trials to breath holds
            design_df = self.assign_trials_to_breath_holds(design_df, timestamps_df)
            
            # Step 4: Save design to CSV
            design_path = os.path.join(self.experiment_log_dir, f"participant_{participant_id}_design.csv")
            design_df.to_csv(design_path, index=False)
            results['files']['design'] = design_path
            
            # Step 5: Generate audio files
            looming_path, tactile_path, log_path = self.generate_participant_audio(
                participant_id, design_df, base_audio_path
            )
            
            if looming_path and tactile_path and log_path:
                results['files']['looming_audio'] = looming_path
                results['files']['tactile_audio'] = tactile_path
                results['files']['stimulus_log'] = log_path
            else:
                error_msg = "Failed to generate audio files"
                results['errors'].append(error_msg)
                print(f"Error: {error_msg}")
            
            # Step 6: Verify generated files
            if log_path:
                verification_results = self.verify_audio_synchronization(participant_id, design_df, log_path)
                if verification_results.get('warnings'):
                    print(f"Warning: {len(verification_results['warnings'])} synchronization issues detected")
            
            # Overall status
            if not results['errors']:
                results['status'] = 'success'
                print(f"=== Successfully processed participant {participant_id} ===\n")
            else:
                print(f"=== Completed processing for participant {participant_id} with errors ===\n")
            
            return results
            
        except Exception as e:
            error_msg = f"Error processing participant {participant_id}: {str(e)}"
            results['errors'].append(error_msg)
            print(error_msg)
            traceback.print_exc()
            return results
    
    def run(self, participant_ids=None, num_participants=None):
        """Run the full audio generation process for multiple participants."""
        from itertools import product  # Import here to avoid circular import issues
        
        if participant_ids is None and num_participants is None:
            num_participants = 5  # Default
        
        if participant_ids is None:
            participant_ids = list(range(num_participants))
        
        results = {}
        
        print(f"=== Starting PPSAudioGenerator ===")
        print(f"Repetitions per condition: {self.config['repetitions_per_condition']}")
        print(f"Processing {len(participant_ids)} participants")
        print(f"Output directories:")
        print(f"  Design files: {self.experiment_log_dir}")
        print(f"  Audio files: {self.audio_output_dir}")
        
        # Ensure we have calculated trial counts
        total_trials = sum(self.trial_counts.values())
        print(f"Total trials per participant: {total_trials}")
        
        # Process each participant
        for p_id in participant_ids:
            results[p_id] = self.process_participant(p_id)
        
        # Generate summary
        success_count = sum(1 for r in results.values() if r['status'] == 'success')
        print(f"\n=== Processing Summary ===")
        print(f"Successfully processed {success_count} of {len(participant_ids)} participants")
        
        # Save overall validation report
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        report_path = os.path.join(self.experiment_log_dir, f"generation_report_{timestamp}.json")
        
        try:
            with open(report_path, 'w') as f:
                # Create a simplified version of results for the report
                report_data = {
                    'timestamp': timestamp,
                    'participants': {
                        p_id: {
                            'status': r['status'],
                            'files': {k: os.path.basename(v) for k, v in r.get('files', {}).items()}
                        } for p_id, r in results.items()
                    }
                }
                json.dump(report_data, f, indent=2)
            
            print(f"Saved generation report to {report_path}")
        except Exception as e:
            print(f"Error saving generation report: {e}")
        
        print(f"=== Completed audio generation ===")
        
        return results

# ======================================================================
# MAIN FUNCTION
# ======================================================================

def main():
    """Main function to run the PPS Audio Generator."""
    # Create configuration
    config = {
        'repetitions_per_condition': 4,
        'included_conditions': {
            'inhalation': True,
            'exhalation': True,
            'baseline': True
        },
        'catch_trial_percentage': 10,
        'debug_mode': True
    }
    
    # Create generator with custom configuration
    generator = PPSAudioGenerator(config)
    
    # Process participants (can specify participant_ids=[0, 1, 2, 3, 4] for specific IDs)
    results = generator.run(num_participants=5)
    
    # Print results
    for p_id, data in results.items():
        status = data['status']
        if status == 'success':
            print(f"Participant {p_id}: Success")
            for file_type, file_path in data.get('files', {}).items():
                print(f"  {file_type}: {os.path.basename(file_path)}")
        else:
            print(f"Participant {p_id}: Error - {', '.join(data.get('errors', ['Unknown error']))}")

if __name__ == "__main__":
    main()
```

```python
import os
import numpy as np
import pandas as pd
import soundfile as sf
import re
from datetime import datetime
import random
import traceback
import json
from itertools import product  # Needed for generating trial combinations

# ======================================================================
# CONFIGURATION PARAMETERS
# ======================================================================

CONFIG = {
    'repetitions_per_condition': 4,
    'included_conditions': {
        'inhalation': True,
        'exhalation': True,
        'baseline': True,
    },
    'catch_trial_percentage': 10,
    'soa_conditions_ms': [190, 400, 700, 1000, 1500],
    'jitter_options_ms': [100, 200, 300, 400, 500],
    'debug_mode': True,
    'audio_extension': {
        'intro_segment_end': '02:00.2',
        'repeat_segment_start': '02:00.2',
        'repeat_segment_end': '04:56.4',
        'preserve_ending': True,
        'ending_duration_sec': 45,
    },
    'paths': {
        'base_dir': r"C:\Users\cogpsy-vrlab\Documents\GitHub\BreathingSpace",
        'level0_dir': r"C:\Users\cogpsy-vrlab\Documents\GitHub\BreathingSpace\Level0_StimulusFiles",
        'experiment_log_dir': r"C:\Users\cogpsy-vrlab\Documents\GitHub\BreathingSpace\Level1_AudioGeneration\ExperimentLog",
        'audio_output_dir': r"C:\Users\cogpsy-vrlab\Documents\GitHub\BreathingSpace\Level1_AudioGeneration\ExperimentAudio",
    }
}

# Updated looming stimuli paths
LOOMING_STIMULI = {
    'blue': "front_az0_FABIAN_HRIR_blue.wav",
    'brown': "front_az0_FABIAN_HRIR_brown.wav",
    'pink': "front_az0_FABIAN_HRIR_pink.wav",
    'white': "front_az0_FABIAN_HRIR_white.wav"
}

# ======================================================================
# PPS AUDIO GENERATOR
# ======================================================================

class PPSAudioGenerator:
    """Generates audio files for Peripersonal Space (PPS) experiments."""
    
    def __init__(self, config=None):
        """Initialize with configuration parameters."""
        self.config = config or CONFIG
        self.setup_paths()
        self.trial_counts = self._calculate_trial_counts()
        self.conditions = [cond for cond, include in self.config['included_conditions'].items() if include]
        
        # Initialize properties
        self.timestamps_df = None
        self.extended_audio = None
        self.extended_audio_path = None
        self.extended_timestamps_df = None
        self.sample_rate = None
        self.looming_stimuli = {}
        self.tactile_stimulus = None
        
        if self.config['debug_mode']:
            print(f"Initialized PPSAudioGenerator with:")
            print(f"  - Repetitions: {self.config['repetitions_per_condition']}")
            print(f"  - Conditions: {self.conditions}")
            print(f"  - SOA values: {self.config['soa_conditions_ms']}")
            print(f"  - Total trials per participant: {sum(self.trial_counts.values())}")
    
    def setup_paths(self):
        """Set up file paths for input and output files."""
        base_dir = self.config['paths']['base_dir']
        level0_dir = self.config['paths']['level0_dir']
        self.experiment_log_dir = self.config['paths']['experiment_log_dir']
        self.audio_output_dir = self.config['paths']['audio_output_dir']
        
        # Input files
        self.breathing_audio_path = os.path.join(level0_dir, "BoxBreathing.wav")
        self.timestamps_csv_path = os.path.join(level0_dir, "BoxBreathing_timestamps.csv")
        
        # Stimulus files
        self.looming_stimuli_paths = {}
        for stim_type, filename in LOOMING_STIMULI.items():
            self.looming_stimuli_paths[stim_type] = os.path.join(level0_dir, filename)
        
        self.tactile_stimulus_path = os.path.join(level0_dir, "tactile_stimulus.wav")
        
        # Create output directories
        os.makedirs(self.experiment_log_dir, exist_ok=True)
        os.makedirs(self.audio_output_dir, exist_ok=True)
        
        if self.config['debug_mode']:
            print(f"Set up paths:")
            print(f"  - Breathing audio: {self.breathing_audio_path}")
            print(f"  - Timestamps CSV: {self.timestamps_csv_path}")
            print(f"  - Looming stimuli: {len(self.looming_stimuli_paths)} files")
            print(f"  - Tactile stimulus: {self.tactile_stimulus_path}")
            print(f"  - Output directories created")
    
    def _calculate_trial_counts(self):
        """Calculate trial counts based on the configuration."""
        reps = self.config['repetitions_per_condition']
        soa_count = len(self.config['soa_conditions_ms'])
        stimulus_types = list(LOOMING_STIMULI.keys())
        catch_percentage = self.config['catch_trial_percentage']
        
        trial_counts = {}
        for condition, include in self.config['included_conditions'].items():
            if include:
                trial_counts[condition] = len(stimulus_types) * soa_count * reps
        
        regular_trials = sum(trial_counts.values())
        catch_trials = int(np.ceil(regular_trials * catch_percentage / 100))
        trial_counts['catch'] = catch_trials
        
        if self.config['debug_mode']:
            print("Trial count calculation:")
            for cond, count in trial_counts.items():
                print(f"  {cond}: {count} trials")
            print(f"  Total: {sum(trial_counts.values())} trials")
        
        return trial_counts
    
    def parse_timestamps(self):
        """Parse the box breathing timestamps CSV."""
        if self.timestamps_df is not None:
            return self.timestamps_df
        
        try:
            df = pd.read_csv(self.timestamps_csv_path)
            
            if 'breathphase' not in df.columns:
                df['breathphase'] = ['inhale' if i % 2 == 0 else 'exhale' for i in range(len(df))]
            
            df['time_seconds'] = df['milliseconds'] / 1000
            self.timestamps_df = df
            
            if self.config['debug_mode']:
                inhale_count = sum(df['breathphase'] == 'inhale')
                exhale_count = sum(df['breathphase'] == 'exhale')
                print(f"Parsed timestamps: {len(df)} total entries")
                print(f"Breath phases: {inhale_count} inhale, {exhale_count} exhale")
            
            return df
            
        except Exception as e:
            print(f"Error parsing timestamps: {e}")
            traceback.print_exc()
            return pd.DataFrame()
    
    def analyze_audio_extension_needs(self):
        """Analyze whether audio extension is needed based on trial count."""
        timestamps_df = self.parse_timestamps()
        total_trials = sum(self.trial_counts.values())
        required_breath_holds = total_trials - self.trial_counts.get('catch', 0)
        
        intro_end = self.config['audio_extension']['intro_segment_end']
        repeat_start = self.config['audio_extension']['repeat_segment_start']
        repeat_end = self.config['audio_extension']['repeat_segment_end']
        
        intro_match = timestamps_df[timestamps_df['timestamp'] == intro_end]
        repeat_start_match = timestamps_df[timestamps_df['timestamp'] == repeat_start]
        repeat_end_match = timestamps_df[timestamps_df['timestamp'] == repeat_end]
        
        intro_end_idx = intro_match.index[0] if not intro_match.empty else None
        repeat_start_idx = repeat_start_match.index[0] if not repeat_start_match.empty else None
        repeat_end_idx = repeat_end_match.index[0] if not repeat_end_match.empty else None
        
        if intro_end_idx is None or repeat_start_idx is None or repeat_end_idx is None:
            total_timestamps = len(timestamps_df)
            intro_end_idx = int(total_timestamps * 0.1)
            repeat_start_idx = intro_end_idx
            repeat_end_idx = int(total_timestamps * 0.8)
        
        available_breath_holds = len(timestamps_df) - (intro_end_idx + 1)
        extension_needed = required_breath_holds > available_breath_holds
        
        if self.config['debug_mode']:
            print("Audio extension analysis:")
            print(f"  Required breath holds: {required_breath_holds}")
            print(f"  Available breath holds: {available_breath_holds}")
            print(f"  Extension needed: {extension_needed}")
            print(f"  Segmentation indices: intro_end={intro_end_idx}, repeat_start={repeat_start_idx}, repeat_end={repeat_end_idx}")
        
        return (extension_needed, required_breath_holds, available_breath_holds, (intro_end_idx, repeat_start_idx, repeat_end_idx))
    
    def extend_breathing_audio(self):
        """Extend the box breathing audio to accommodate all trials."""
        extension_info = self.analyze_audio_extension_needs()
        extension_needed, required_holds, available_holds, segment_indices = extension_info
        intro_end_idx, repeat_start_idx, repeat_end_idx = segment_indices
        
        timestamps_df = self.parse_timestamps()
        
        if not extension_needed:
            if self.config['debug_mode']:
                print("Audio extension not needed, using original audio file.")
            return self.breathing_audio_path, timestamps_df
        
        original_audio, sample_rate = sf.read(self.breathing_audio_path)
        self.sample_rate = sample_rate
        
        intro_end_sample = int(timestamps_df.iloc[intro_end_idx]['sample_index'])
        repeat_start_sample = int(timestamps_df.iloc[repeat_start_idx]['sample_index'])
        repeat_end_sample = int(timestamps_df.iloc[repeat_end_idx]['sample_index'])
        
        segment_size = repeat_end_idx - repeat_start_idx + 1
        segment_samples = repeat_end_sample - repeat_start_sample
        
        additional_holds_needed = required_holds - available_holds
        additional_segments_needed = int(np.ceil(additional_holds_needed / segment_size))
        
        if self.config['debug_mode']:
            print("Audio extension details:")
            print(f"  Segment size: {segment_size} breath holds ({segment_samples} samples)")
            print(f"  Additional holds needed: {additional_holds_needed}")
            print(f"  Additional segments needed: {additional_segments_needed}")
        
        extended_audio = original_audio[:intro_end_sample].copy()
        
        for i in range(additional_segments_needed + 1):
            segment = original_audio[repeat_start_sample:repeat_end_sample].copy()
            if len(extended_audio.shape) > 1 and len(segment.shape) > 1:
                extended_audio = np.vstack([extended_audio, segment])
            else:
                if len(extended_audio.shape) > 1:
                    segment_stereo = np.column_stack([segment, segment])
                    extended_audio = np.vstack([extended_audio, segment_stereo])
                elif len(segment.shape) > 1:
                    extended_mono = np.column_stack([extended_audio, extended_audio])
                    segment_mono = np.mean(segment, axis=1)
                    extended_audio = np.vstack([extended_mono, segment_mono])
                else:
                    extended_audio = np.concatenate([extended_audio, segment])
        
        if self.config['audio_extension']['preserve_ending']:
            ending_duration_sec = self.config['audio_extension']['ending_duration_sec']
            ending_samples = min(
                len(original_audio) - repeat_end_sample,
                int(ending_duration_sec * sample_rate)
            )
            
            ending_segment = original_audio[repeat_end_sample:repeat_end_sample+ending_samples].copy()
            
            if len(extended_audio.shape) > 1 and len(ending_segment.shape) > 1:
                extended_audio = np.vstack([extended_audio, ending_segment])
            else:
                if len(extended_audio.shape) > 1:
                    ending_stereo = np.column_stack([ending_segment, ending_segment])
                    extended_audio = np.vstack([extended_audio, ending_stereo])
                elif len(ending_segment.shape) > 1:
                    extended_mono = np.column_stack([extended_audio, extended_audio])
                    ending_mono = np.mean(ending_segment, axis=1)
                    extended_audio = np.vstack([extended_mono, ending_mono])
                else:
                    extended_audio = np.concatenate([extended_audio, ending_segment])
        
        extended_timestamps = timestamps_df.iloc[:intro_end_idx+1].copy()
        time_offset = 0
        sample_offset = 0
        milliseconds_offset = 0
        
        for i in range(additional_segments_needed + 1):
            segment_timestamps = timestamps_df.iloc[repeat_start_idx:repeat_end_idx+1].copy()
            
            if i > 0:
                segment_duration_sec = (timestamps_df.iloc[repeat_end_idx]['time_seconds'] - 
                                          timestamps_df.iloc[repeat_start_idx]['time_seconds'])
                segment_milliseconds = (timestamps_df.iloc[repeat_end_idx]['milliseconds'] - 
                                        timestamps_df.iloc[repeat_start_idx]['milliseconds'])
                
                time_offset += segment_duration_sec
                milliseconds_offset += segment_milliseconds
                sample_offset += segment_samples
                
                segment_timestamps['time_seconds'] += time_offset
                segment_timestamps['milliseconds'] += milliseconds_offset
                segment_timestamps['sample_index'] += sample_offset
                
                for idx, row in segment_timestamps.iterrows():
                    minutes = int(row['time_seconds'] // 60)
                    seconds = row['time_seconds'] % 60
                    segment_timestamps.at[idx, 'timestamp'] = f"{minutes:02d}:{seconds:.1f}"
            
            extended_timestamps = pd.concat([extended_timestamps, segment_timestamps])
        
        if self.config['audio_extension']['preserve_ending']:
            remaining_timestamps = timestamps_df.iloc[repeat_end_idx+1:].copy()
            
            if not remaining_timestamps.empty:
                remaining_timestamps['time_seconds'] += time_offset
                remaining_timestamps['milliseconds'] += milliseconds_offset
                remaining_timestamps['sample_index'] += sample_offset
                
                for idx, row in remaining_timestamps.iterrows():
                    minutes = int(row['time_seconds'] // 60)
                    seconds = row['time_seconds'] % 60
                    remaining_timestamps.at[idx, 'timestamp'] = f"{minutes:02d}:{seconds:.1f}"
                
                extended_timestamps = pd.concat([extended_timestamps, remaining_timestamps])
        
        extended_timestamps = extended_timestamps.reset_index(drop=True)
        
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        extended_audio_path = os.path.join(
            self.audio_output_dir, 
            f"extended_boxbreathing_{timestamp}.wav"
        )
        sf.write(extended_audio_path, extended_audio, sample_rate)
        
        extended_timestamps_path = os.path.join(
            self.experiment_log_dir, 
            f"extended_boxbreathing_timestamps_{timestamp}.csv"
        )
        extended_timestamps.to_csv(extended_timestamps_path, index=False)
        
        if self.config['debug_mode']:
            print("Extended audio created:")
            print(f"  Original length: {len(original_audio)/sample_rate:.2f} seconds")
            print(f"  Extended length: {len(extended_audio)/sample_rate:.2f} seconds")
            print(f"  Extended audio saved to: {extended_audio_path}")
            print(f"  Extended timestamps saved to: {extended_timestamps_path}")
        
        self.extended_audio = extended_audio
        self.extended_audio_path = extended_audio_path
        self.extended_timestamps_df = extended_timestamps
        
        return extended_audio_path, extended_timestamps
    
    def load_stimuli(self):
        """Load all stimulus files."""
        if self.looming_stimuli and self.tactile_stimulus is not None:
            return self.looming_stimuli, self.tactile_stimulus
        
        looming_stimuli = {}
        for stim_type, path in self.looming_stimuli_paths.items():
            try:
                data, sr = sf.read(path)
                if self.sample_rate is None:
                    self.sample_rate = sr
                elif sr != self.sample_rate:
                    print(f"Warning: Sample rate mismatch for {stim_type}: {sr} vs {self.sample_rate}")
                looming_stimuli[stim_type] = data
                if self.config['debug_mode']:
                    print(f"Loaded {stim_type} looming stimulus: {len(data)} samples")
            except Exception as e:
                print(f"Error loading {stim_type} stimulus: {e}")
        
        try:
            tactile_data, sr = sf.read(self.tactile_stimulus_path)
            if self.sample_rate is None:
                self.sample_rate = sr
            elif sr != self.sample_rate:
                print(f"Warning: Sample rate mismatch for tactile: {sr} vs {self.sample_rate}")
            self.tactile_stimulus = tactile_data
            if self.config['debug_mode']:
                print(f"Loaded tactile stimulus: {len(tactile_data)} samples")
        except Exception as e:
            print(f"Error loading tactile stimulus: {e}")
            self.tactile_stimulus = None
        
        self.looming_stimuli = looming_stimuli
        return looming_stimuli, self.tactile_stimulus
    
    def generate_counterbalanced_design(self, participant_id):
        """Generate a counterbalanced design for a participant."""
        soa_conditions = self.config['soa_conditions_ms']
        jitter_options = self.config['jitter_options_ms']
        stimulus_types = list(LOOMING_STIMULI.keys())
        
        all_trials = []
        
        for condition, count in self.trial_counts.items():
            if condition == 'catch' or count == 0:
                continue
            
            stim_soa_combinations = list(product(stimulus_types, soa_conditions))
            reps_needed = count // len(stim_soa_combinations)
            remainder = count % len(stim_soa_combinations)
            
            full_combinations = stim_soa_combinations * reps_needed
            remainder_combinations = random.sample(stim_soa_combinations, remainder)
            all_combinations = full_combinations + remainder_combinations
            random.shuffle(all_combinations)
            
            for i, (stim_type, soa) in enumerate(all_combinations):
                all_trials.append({
                    'participant_id': participant_id,
                    'trial_number': len(all_trials) + 1,
                    'stimulus_type': stim_type,
                    'trial_type': condition,
                    'soa_condition_idx': soa_conditions.index(soa),
                    'soa_value_ms': soa,
                    'jitter_ms': random.choice(jitter_options),
                    'is_tactile': True
                })
        
        catch_count = self.trial_counts.get('catch', 0)
        if catch_count > 0:
            catch_stim_combinations = list(product(stimulus_types, soa_conditions))
            while len(catch_stim_combinations) < catch_count:
                catch_stim_combinations.extend(catch_stim_combinations)
            catch_combinations = random.sample(catch_stim_combinations, catch_count)
            
            for i, (stim_type, soa) in enumerate(catch_combinations):
                all_trials.append({
                    'participant_id': participant_id,
                    'trial_number': len(all_trials) + 1,
                    'stimulus_type': stim_type,
                    'trial_type': 'catch',
                    'soa_condition_idx': soa_conditions.index(soa),
                    'soa_value_ms': soa,
                    'jitter_ms': random.choice(jitter_options),
                    'is_tactile': False
                })
        
        design_df = pd.DataFrame(all_trials)
        design_df = design_df.sample(frac=1, random_state=participant_id*100).reset_index(drop=True)
        design_df['trial_number'] = range(1, len(design_df) + 1)
        return design_df
    
    def assign_trials_to_breath_holds(self, design_df, timestamps_df):
        """Assign trials to specific breath holds while maintaining balance."""
        intro_end = self.config['audio_extension']['intro_segment_end']
        intro_end_match = timestamps_df[timestamps_df['timestamp'] == intro_end]
        intro_end_idx = intro_end_match.index[0] if not intro_end_match.empty else int(len(timestamps_df) * 0.1)
        if intro_end_match.empty:
            print(f"Warning: Using fallback intro_end_idx = {intro_end_idx}")
        
        usable_timestamps = timestamps_df.iloc[intro_end_idx+1:].copy()
        inhale_timestamps = usable_timestamps[usable_timestamps['breathphase'] == 'inhale']
        exhale_timestamps = usable_timestamps[usable_timestamps['breathphase'] == 'exhale']
        
        if self.config['debug_mode']:
            print("Assigning trials to breath holds:")
            print(f"  Available inhale holds: {len(inhale_timestamps)}")
            print(f"  Available exhale holds: {len(exhale_timestamps)}")
        
        inhale_indices = inhale_timestamps.index.tolist()
        exhale_indices = exhale_timestamps.index.tolist()
        trial_assignments = {}
        
        for condition in set(design_df['trial_type']):
            condition_trials = design_df[design_df['trial_type'] == condition]
            condition_count = len(condition_trials)
            inhale_count = condition_count // 2
            exhale_count = condition_count - inhale_count
            
            if self.config['debug_mode']:
                print(f"  Condition '{condition}': {condition_count} trials - {inhale_count} inhale, {exhale_count} exhale")
            
            if inhale_count > len(inhale_indices) or exhale_count > len(exhale_indices):
                print(f"Warning: Not enough breath holds for condition '{condition}'")
                print(f"  Need {inhale_count} inhale holds, have {len(inhale_indices)}")
                print(f"  Need {exhale_count} exhale holds, have {len(exhale_indices)}")
                inhale_count = min(inhale_count, len(inhale_indices))
                exhale_count = min(exhale_count, len(exhale_indices))
            
            for idx, trial_idx in zip(inhale_indices[:inhale_count], condition_trials.index[:inhale_count]):
                timestamp_row = timestamps_df.loc[idx]
                trial_assignments[trial_idx] = {
                    'timestamp_idx': idx,
                    'breathphase': 'inhale',
                    'timestamp': timestamp_row['timestamp'],
                    'milliseconds': timestamp_row['milliseconds'],
                    'sample_index': timestamp_row['sample_index'],
                    'time_seconds': timestamp_row['time_seconds']
                }
            
            for idx, trial_idx in zip(exhale_indices[:exhale_count], condition_trials.index[inhale_count:inhale_count+exhale_count]):
                timestamp_row = timestamps_df.loc[idx]
                trial_assignments[trial_idx] = {
                    'timestamp_idx': idx,
                    'breathphase': 'exhale',
                    'timestamp': timestamp_row['timestamp'],
                    'milliseconds': timestamp_row['milliseconds'],
                    'sample_index': timestamp_row['sample_index'],
                    'time_seconds': timestamp_row['time_seconds'] 
                }
            
            inhale_indices = inhale_indices[inhale_count:]
            exhale_indices = exhale_indices[exhale_count:]
        
        design_df['breathphase'] = design_df.index.map(lambda idx: trial_assignments.get(idx, {}).get('breathphase'))
        design_df['timestamp'] = design_df.index.map(lambda idx: trial_assignments.get(idx, {}).get('timestamp'))
        design_df['milliseconds'] = design_df.index.map(lambda idx: trial_assignments.get(idx, {}).get('milliseconds'))
        design_df['sample_index'] = design_df.index.map(lambda idx: trial_assignments.get(idx, {}).get('sample_index'))
        design_df['time_seconds'] = design_df.index.map(lambda idx: trial_assignments.get(idx, {}).get('time_seconds'))
        
        unassigned = design_df[design_df['breathphase'].isna()]
        if not unassigned.empty:
            print(f"Warning: {len(unassigned)} trials could not be assigned to breath holds")
            available_timestamps = timestamps_df.loc[
                ~timestamps_df.index.isin([a['timestamp_idx'] for a in trial_assignments.values() if 'timestamp_idx' in a])
            ]
            if not available_timestamps.empty:
                print(f"  Using fallback assignment with {len(available_timestamps)} available timestamps")
                for idx in unassigned.index:
                    if available_timestamps.empty:
                        break
                    ts_idx = random.choice(available_timestamps.index)
                    timestamp_row = available_timestamps.loc[ts_idx]
                    design_df.at[idx, 'breathphase'] = timestamp_row.get('breathphase', 'unknown')
                    design_df.at[idx, 'timestamp'] = timestamp_row['timestamp']
                    design_df.at[idx, 'milliseconds'] = timestamp_row['milliseconds']
                    design_df.at[idx, 'sample_index'] = timestamp_row['sample_index']
                    design_df.at[idx, 'time_seconds'] = timestamp_row['time_seconds']
                    available_timestamps = available_timestamps.drop(ts_idx)
        
        design_df['jittered_ms'] = design_df.apply(
            lambda row: row['milliseconds'] + row['jitter_ms'] if pd.notnull(row['milliseconds']) else None,
            axis=1
        )
        
        design_df['soa_ms'] = design_df.apply(
            lambda row: row['jittered_ms'] + row['soa_value_ms'] if pd.notnull(row['jittered_ms']) and row['trial_type'] != 'catch' else None, 
            axis=1
        )
        
        design_df['timestamp_original'] = design_df.apply(
            lambda row: f"{int(row['milliseconds'] // 60000):02}:{(row['milliseconds'] % 60000) / 1000:.1f}" if pd.notnull(row['milliseconds']) else None,
            axis=1
        )
        
        design_df['timestamp_after_jitter'] = design_df.apply(
            lambda row: f"{int(row['jittered_ms'] // 60000):02}:{(row['jittered_ms'] % 60000) / 1000:.1f}" if pd.notnull(row['jittered_ms']) else None,
            axis=1
        )
        
        design_df['timestamp_with_soa'] = design_df.apply(
            lambda row: f"{int(row['soa_ms'] // 60000):02}:{(row['soa_ms'] % 60000) / 1000:.1f}" if pd.notnull(row['soa_ms']) else None,
            axis=1
        )
        
        sample_rate = self.sample_rate if self.sample_rate else 48000
        design_df['base_ms'] = design_df['milliseconds']
        design_df['jittered_sample'] = design_df.apply(
            lambda row: int(row['jittered_ms'] / 1000 * sample_rate) if pd.notnull(row['jittered_ms']) else None,
            axis=1
        )
        design_df['soa_sample'] = design_df.apply(
            lambda row: int(row['soa_ms'] / 1000 * sample_rate) if pd.notnull(row['soa_ms']) else None,
            axis=1
        )
        
        return design_df
    
    def generate_participant_audio(self, participant_id, design_df, base_audio_path):
        """Generate participant-specific audio files."""
        try:
            base_audio, sample_rate = sf.read(base_audio_path)
            if self.sample_rate is None:
                self.sample_rate = sample_rate
        except Exception as e:
            print(f"Error loading base audio: {e}")
            return None, None, None
        
        looming_stimuli, tactile_stimulus = self.load_stimuli()
        
        looming_audio = base_audio.copy()
        
        if len(base_audio.shape) > 1:
            tactile_audio = np.zeros((len(base_audio), base_audio.shape[1]), dtype=base_audio.dtype)
        else:
            tactile_audio = np.zeros(len(base_audio), dtype=base_audio.dtype)
        
        if tactile_stimulus is not None:
            if len(tactile_stimulus.shape) == 1 and len(tactile_audio.shape) > 1:
                tactile_stereo = np.column_stack((tactile_stimulus, tactile_stimulus))
            elif len(tactile_stimulus.shape) > 1 and len(tactile_audio.shape) == 1:
                tactile_stereo = np.mean(tactile_stimulus, axis=1)
            else:
                tactile_stereo = tactile_stimulus
        else:
            print("Warning: No tactile stimulus available")
            tactile_stereo = None
        
        injection_log = []
        
        for _, row in design_df.iterrows():
            if pd.isnull(row['jittered_sample']):
                continue
            
            looming_pos = int(row['jittered_sample'])
            
            if row['trial_type'] != 'catch' and pd.notnull(row['soa_sample']):
                tactile_pos = int(row['soa_sample'])
            else:
                tactile_pos = None
            
            stim_type = row['stimulus_type']
            if stim_type in looming_stimuli:
                looming_data = looming_stimuli[stim_type]
                if looming_pos + len(looming_data) <= len(looming_audio):
                    if len(looming_data.shape) == 2 and len(looming_audio.shape) == 2:
                        looming_audio[looming_pos:looming_pos+len(looming_data)] += looming_data
                    elif len(looming_data.shape) == 1 and len(looming_audio.shape) == 2:
                        for ch in range(looming_audio.shape[1]):
                            looming_audio[looming_pos:looming_pos+len(looming_data), ch] += looming_data
                    elif len(looming_data.shape) == 2 and len(looming_audio.shape) == 1:
                        looming_audio[looming_pos:looming_pos+len(looming_data)] += np.mean(looming_data, axis=1)
                    else:
                        looming_audio[looming_pos:looming_pos+len(looming_data)] += looming_data
                    
                    injection_log.append({
                        'trial_number': row['trial_number'],
                        'trial_type': row['trial_type'],
                        'stimulus_type': stim_type,
                        'breathphase': row.get('breathphase', 'unknown'),
                        'looming_position': looming_pos,
                        'looming_time_ms': row['jittered_ms']
                    })
            
            if tactile_pos is not None and tactile_stereo is not None:
                if tactile_pos + len(tactile_stereo) <= len(tactile_audio):
                    if len(tactile_stereo.shape) == 2 and len(tactile_audio.shape) == 2:
                        tactile_audio[tactile_pos:tactile_pos+len(tactile_stereo)] += tactile_stereo
                    elif len(tactile_stereo.shape) == 1 and len(tactile_audio.shape) == 2:
                        for ch in range(tactile_audio.shape[1]):
                            tactile_audio[tactile_pos:tactile_pos+len(tactile_stereo), ch] += tactile_stereo
                    elif len(tactile_stereo.shape) == 2 and len(tactile_audio.shape) == 1:
                        tactile_audio[tactile_pos:tactile_pos+len(tactile_stereo)] += np.mean(tactile_stereo, axis=1)
                    else:
                        tactile_audio[tactile_pos:tactile_pos+len(tactile_stereo)] += tactile_stereo
                    
                    for log_entry in injection_log:
                        if log_entry['trial_number'] == row['trial_number']:
                            log_entry['tactile_position'] = tactile_pos
                            log_entry['tactile_time_ms'] = row['soa_ms']
                            log_entry['soa_ms'] = row['soa_value_ms']
                            break
        
        looming_filename = f"participant_{participant_id}_design_looming.wav"
        tactile_filename = f"participant_{participant_id}_design_tactile.wav"
        log_filename = f"participant_{participant_id}_design_stimuli_log.csv"
        
        looming_path = os.path.join(self.audio_output_dir, looming_filename)
        tactile_path = os.path.join(self.audio_output_dir, tactile_filename)
        log_path = os.path.join(self.audio_output_dir, log_filename)
        
        try:
            sf.write(looming_path, looming_audio, sample_rate)
            sf.write(tactile_path, tactile_audio, sample_rate)
            
            injection_df = pd.DataFrame(injection_log)
            injection_df.to_csv(log_path, index=False)
            
            if self.config['debug_mode']:
                print(f"Generated audio files for participant {participant_id}:")
                print(f"  Looming audio: {looming_path}")
                print(f"  Tactile audio: {tactile_path}")
                print(f"  Injection log: {log_path}")
                print(f"  Injected {len(injection_log)} stimuli")
            
            return looming_path, tactile_path, log_path
        except Exception as e:
            print(f"Error saving audio files: {e}")
            return None, None, None
    
    def verify_audio_synchronization(self, participant_id, design_df, injection_log_path):
        """Verify the synchronization between generated audio files and design."""
        results = {
            'success': False,
            'error': None,
            'warnings': []
        }
        
        try:
            injection_df = pd.read_csv(injection_log_path)
            design_count = len(design_df)
            injection_count = len(injection_df)
            
            if design_count != injection_count:
                warning = f"Design has {design_count} trials but injection log has {injection_count} trials"
                results['warnings'].append(warning)
            
            sample_size = min(5, len(injection_df))
            sample_trials = injection_df.sample(n=sample_size) if len(injection_df) > 5 else injection_df
            
            for _, row in sample_trials.iterrows():
                trial_number = row['trial_number']
                design_row = design_df[design_df['trial_number'] == trial_number]
                
                if len(design_row) == 0:
                    results['warnings'].append(f"Trial {trial_number} not found in design")
                    continue
                
                design_row = design_row.iloc[0]
                
                if 'looming_time_ms' in row and 'jittered_ms' in design_row:
                    looming_diff = abs(row['looming_time_ms'] - design_row['jittered_ms'])
                    if looming_diff > 1:
                        results['warnings'].append(f"Trial {trial_number}: Looming timing mismatch: {looming_diff} ms")
                
                if design_row['trial_type'] != 'catch' and 'tactile_time_ms' in row and 'soa_ms' in design_row:
                    tactile_diff = abs(row['tactile_time_ms'] - design_row['soa_ms'])
                    if tactile_diff > 1:
                        results['warnings'].append(f"Trial {trial_number}: Tactile timing mismatch: {tactile_diff} ms")
            
            results['success'] = True
            
            if self.config['debug_mode']:
                print(f"Verification complete for participant {participant_id}")
                if results['warnings']:
                    print(f"  Warnings: {len(results['warnings'])}")
                    for warning in results['warnings'][:3]:
                        print(f"    - {warning}")
                    if len(results['warnings']) > 3:
                        print(f"    - (and {len(results['warnings']) - 3} more warnings)")
                else:
                    print("  No warnings - all timings match expected values")
            
            return results
            
        except Exception as e:
            results['error'] = f"Error during verification: {str(e)}"
            results['success'] = False
            if self.config['debug_mode']:
                print(results['error'])
            return results
    
    def process_participant(self, participant_id):
        """Process a single participant from design generation to audio creation."""
        results = {
            'participant_id': participant_id,
            'status': 'error',
            'files': {},
            'errors': []
        }
        
        try:
            print(f"\n=== Processing participant {participant_id} ===")
            
            if self.extended_audio_path is None:
                base_audio_path, timestamps_df = self.extend_breathing_audio()
            else:
                base_audio_path = self.extended_audio_path
                timestamps_df = self.extended_timestamps_df
            
            design_df = self.generate_counterbalanced_design(participant_id)
            design_df = self.assign_trials_to_breath_holds(design_df, timestamps_df)
            
            design_path = os.path.join(self.experiment_log_dir, f"participant_{participant_id}_design.csv")
            design_df.to_csv(design_path, index=False)
            results['files']['design'] = design_path
            
            looming_path, tactile_path, log_path = self.generate_participant_audio(
                participant_id, design_df, base_audio_path
            )
            
            if looming_path and tactile_path and log_path:
                results['files']['looming_audio'] = looming_path
                results['files']['tactile_audio'] = tactile_path
                results['files']['stimulus_log'] = log_path
            else:
                error_msg = "Failed to generate audio files"
                results['errors'].append(error_msg)
                print(f"Error: {error_msg}")
            
            if log_path:
                verification_results = self.verify_audio_synchronization(participant_id, design_df, log_path)
                if verification_results.get('warnings'):
                    print(f"Warning: {len(verification_results['warnings'])} synchronization issues detected")
            
            if not results['errors']:
                results['status'] = 'success'
                print(f"=== Successfully processed participant {participant_id} ===\n")
            else:
                print(f"=== Completed processing for participant {participant_id} with errors ===\n")
            
            return results
            
        except Exception as e:
            error_msg = f"Error processing participant {participant_id}: {str(e)}"
            results['errors'].append(error_msg)
            print(error_msg)
            traceback.print_exc()
            return results
    
    def run(self, participant_ids=None, num_participants=None):
        """Run the full audio generation process for multiple participants."""
        if participant_ids is None and num_participants is None:
            num_participants = 5
        
        if participant_ids is None:
            participant_ids = list(range(num_participants))
        
        results = {}
        
        print("=== Starting PPSAudioGenerator ===")
        print(f"Repetitions per condition: {self.config['repetitions_per_condition']}")
        print(f"Processing {len(participant_ids)} participants")
        print("Output directories:")
        print(f"  Design files: {self.experiment_log_dir}")
        print(f"  Audio files: {self.audio_output_dir}")
        total_trials = sum(self.trial_counts.values())
        print(f"Total trials per participant: {total_trials}")
        
        for p_id in participant_ids:
            results[p_id] = self.process_participant(p_id)
        
        success_count = sum(1 for r in results.values() if r['status'] == 'success')
        print("\n=== Processing Summary ===")
        print(f"Successfully processed {success_count} of {len(participant_ids)} participants")
        
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        report_path = os.path.join(self.experiment_log_dir, f"generation_report_{timestamp}.json")
        
        try:
            with open(report_path, 'w') as f:
                report_data = {
                    'timestamp': timestamp,
                    'participants': {
                        p_id: {
                            'status': r['status'],
                            'files': {k: os.path.basename(v) for k, v in r.get('files', {}).items()}
                        } for p_id, r in results.items()
                    }
                }
                json.dump(report_data, f, indent=2)
            print(f"Saved generation report to {report_path}")
        except Exception as e:
            print(f"Error saving generation report: {e}")
        
        print("=== Completed audio generation ===")
        return results

# ======================================================================
# MAIN FUNCTION
# ======================================================================

def main():
    """Main function to run the PPS Audio Generator."""
    generator = PPSAudioGenerator(CONFIG)
    results = generator.run(num_participants=5)
    
    for p_id, data in results.items():
        status = data['status']
        if status == 'success':
            print(f"Participant {p_id}: Success")
            for file_type, file_path in data.get('files', {}).items():
                print(f"  {file_type}: {os.path.basename(file_path)}")
        else:
            print(f"Participant {p_id}: Error - {', '.join(data.get('errors', ['Unknown error']))}")

if __name__ == "__main__":
    main()
```

```python
import os
import numpy as np
import pandas as pd
import soundfile as sf
from datetime import datetime
import random
import traceback
import json
from itertools import product  # For generating trial combinations

# ======================================================================
# CONFIGURATION PARAMETERS
# ======================================================================

CONFIG = {
    # Repetition parameter: how many times each stimulus combination (condition + SOA) is repeated
    'repetitions': 4,
    # Number of participants for the experiment
    'num_participants': 5,
    'included_conditions': {
        'inhalation': True,
        'exhalation': True,
        'baseline': True,
    },
    'catch_trial_percentage': 10,
    'soa_conditions_ms': [190, 400, 700, 1000, 1500],
    'jitter_options_ms': [100, 200, 300, 400, 500],
    'debug_mode': True,
    'audio_extension': {
        'intro_segment_end': '02:00.2',
        'repeat_segment_start': '02:00.2',
        'repeat_segment_end': '04:56.4',
        'preserve_ending': True,
        'ending_duration_sec': 45,
    },
    'paths': {
        'base_dir': r"C:\Users\cogpsy-vrlab\Documents\GitHub\BreathingSpace",
        'level0_dir': r"C:\Users\cogpsy-vrlab\Documents\GitHub\BreathingSpace\Level0_StimulusFiles",
        'experiment_log_dir': r"C:\Users\cogpsy-vrlab\Documents\GitHub\BreathingSpace\Level1_AudioGeneration\ExperimentLog",
        'audio_output_dir': r"C:\Users\cogpsy-vrlab\Documents\GitHub\BreathingSpace\Level1_AudioGeneration\ExperimentAudio",
    }
}

# Updated looming stimuli paths
LOOMING_STIMULI = {
    'blue': "front_az0_FABIAN_HRIR_blue.wav",
    'brown': "front_az0_FABIAN_HRIR_brown.wav",
    'pink': "front_az0_FABIAN_HRIR_pink.wav",
    'white': "front_az0_FABIAN_HRIR_white.wav"
}

# ======================================================================
# PPS AUDIO GENERATOR
# ======================================================================

class PPSAudioGenerator:
    """Generates audio files for Peripersonal Space (PPS) experiments."""
    
    def __init__(self, config=None):
        """Initialize with configuration parameters."""
        self.config = config or CONFIG
        self.setup_paths()
        self.trial_counts = self._calculate_trial_counts()
        self.conditions = [cond for cond, include in self.config['included_conditions'].items() if include]
        
        # Initialize properties
        self.timestamps_df = None
        self.extended_audio = None
        self.extended_audio_path = None
        self.extended_timestamps_df = None
        self.sample_rate = None
        self.looming_stimuli = {}
        self.tactile_stimulus = None
        
        if self.config['debug_mode']:
            print(f"Initialized PPSAudioGenerator with:")
            print(f"  - Repetitions: {self.config['repetitions']}")
            print(f"  - Conditions: {self.conditions}")
            print(f"  - SOA values: {self.config['soa_conditions_ms']}")
            print(f"  - Total trials per participant: {sum(self.trial_counts.values())}")
    
    def setup_paths(self):
        """Set up file paths for input and output files."""
        base_dir = self.config['paths']['base_dir']
        level0_dir = self.config['paths']['level0_dir']
        self.experiment_log_dir = self.config['paths']['experiment_log_dir']
        self.audio_output_dir = self.config['paths']['audio_output_dir']
        
        # Input files
        self.breathing_audio_path = os.path.join(level0_dir, "BoxBreathing.wav")
        self.timestamps_csv_path = os.path.join(level0_dir, "BoxBreathing_timestamps.csv")
        
        # Stimulus files
        self.looming_stimuli_paths = {}
        for stim_type, filename in LOOMING_STIMULI.items():
            self.looming_stimuli_paths[stim_type] = os.path.join(level0_dir, filename)
        
        self.tactile_stimulus_path = os.path.join(level0_dir, "tactile_stimulus.wav")
        
        # Create output directories if they don't exist
        os.makedirs(self.experiment_log_dir, exist_ok=True)
        os.makedirs(self.audio_output_dir, exist_ok=True)
        
        if self.config['debug_mode']:
            print("Set up paths:")
            print(f"  - Breathing audio: {self.breathing_audio_path}")
            print(f"  - Timestamps CSV: {self.timestamps_csv_path}")
            print(f"  - Looming stimuli: {len(self.looming_stimuli_paths)} files")
            print(f"  - Tactile stimulus: {self.tactile_stimulus_path}")
            print("  - Output directories created")
    
    def _calculate_trial_counts(self):
        """Calculate trial counts based on the configuration."""
        reps = self.config['repetitions']
        soa_count = len(self.config['soa_conditions_ms'])
        stimulus_types = list(LOOMING_STIMULI.keys())
        catch_percentage = self.config['catch_trial_percentage']
        
        trial_counts = {}
        for condition, include in self.config['included_conditions'].items():
            if include:
                # Each condition (inhale, exhale, baseline) gets (# noise types) × (# SOA values) × (repetitions)
                trial_counts[condition] = len(stimulus_types) * soa_count * reps
        
        regular_trials = sum(trial_counts.values())
        catch_trials = int(np.ceil(regular_trials * catch_percentage / 100))
        trial_counts['catch'] = catch_trials
        
        if self.config['debug_mode']:
            print("Trial count calculation:")
            for cond, count in trial_counts.items():
                print(f"  {cond}: {count} trials")
            print(f"  Total: {sum(trial_counts.values())} trials")
        
        return trial_counts
    
    def parse_timestamps(self):
        """Parse the box breathing timestamps CSV."""
        if self.timestamps_df is not None:
            return self.timestamps_df
        
        try:
            df = pd.read_csv(self.timestamps_csv_path)
            if 'breathphase' not in df.columns:
                df['breathphase'] = ['inhale' if i % 2 == 0 else 'exhale' for i in range(len(df))]
            df['time_seconds'] = df['milliseconds'] / 1000
            self.timestamps_df = df
            
            if self.config['debug_mode']:
                inhale_count = sum(df['breathphase'] == 'inhale')
                exhale_count = sum(df['breathphase'] == 'exhale')
                print(f"Parsed timestamps: {len(df)} total entries")
                print(f"Breath phases: {inhale_count} inhale, {exhale_count} exhale")
            
            return df
        except Exception as e:
            print(f"Error parsing timestamps: {e}")
            traceback.print_exc()
            return pd.DataFrame()
    
    def analyze_audio_extension_needs(self):
        """Analyze whether audio extension is needed based on trial count."""
        timestamps_df = self.parse_timestamps()
        total_trials = sum(self.trial_counts.values())
        required_breath_holds = total_trials - self.trial_counts.get('catch', 0)
        
        intro_end = self.config['audio_extension']['intro_segment_end']
        repeat_start = self.config['audio_extension']['repeat_segment_start']
        repeat_end = self.config['audio_extension']['repeat_segment_end']
        
        intro_match = timestamps_df[timestamps_df['timestamp'] == intro_end]
        repeat_start_match = timestamps_df[timestamps_df['timestamp'] == repeat_start]
        repeat_end_match = timestamps_df[timestamps_df['timestamp'] == repeat_end]
        
        intro_end_idx = intro_match.index[0] if not intro_match.empty else None
        repeat_start_idx = repeat_start_match.index[0] if not repeat_start_match.empty else None
        repeat_end_idx = repeat_end_match.index[0] if not repeat_end_match.empty else None
        
        if intro_end_idx is None or repeat_start_idx is None or repeat_end_idx is None:
            total_timestamps = len(timestamps_df)
            intro_end_idx = int(total_timestamps * 0.1)
            repeat_start_idx = intro_end_idx
            repeat_end_idx = int(total_timestamps * 0.8)
        
        available_breath_holds = len(timestamps_df) - (intro_end_idx + 1)
        extension_needed = required_breath_holds > available_breath_holds
        
        if self.config['debug_mode']:
            print("Audio extension analysis:")
            print(f"  Required breath holds: {required_breath_holds}")
            print(f"  Available breath holds: {available_breath_holds}")
            print(f"  Extension needed: {extension_needed}")
            print(f"  Segmentation indices: intro_end={intro_end_idx}, repeat_start={repeat_start_idx}, repeat_end={repeat_end_idx}")
        
        return (extension_needed, required_breath_holds, available_breath_holds, (intro_end_idx, repeat_start_idx, repeat_end_idx))
    
    def extend_breathing_audio(self):
        """Extend the box breathing audio to accommodate all trials."""
        extension_info = self.analyze_audio_extension_needs()
        extension_needed, required_holds, available_holds, segment_indices = extension_info
        intro_end_idx, repeat_start_idx, repeat_end_idx = segment_indices
        
        timestamps_df = self.parse_timestamps()
        if not extension_needed:
            if self.config['debug_mode']:
                print("Audio extension not needed, using original audio file.")
            return self.breathing_audio_path, timestamps_df
        
        original_audio, sample_rate = sf.read(self.breathing_audio_path)
        self.sample_rate = sample_rate
        
        intro_end_sample = int(timestamps_df.iloc[intro_end_idx]['sample_index'])
        repeat_start_sample = int(timestamps_df.iloc[repeat_start_idx]['sample_index'])
        repeat_end_sample = int(timestamps_df.iloc[repeat_end_idx]['sample_index'])
        
        segment_size = repeat_end_idx - repeat_start_idx + 1
        segment_samples = repeat_end_sample - repeat_start_sample
        
        additional_holds_needed = required_holds - available_holds
        additional_segments_needed = int(np.ceil(additional_holds_needed / segment_size))
        
        if self.config['debug_mode']:
            print("Audio extension details:")
            print(f"  Segment size: {segment_size} breath holds ({segment_samples} samples)")
            print(f"  Additional holds needed: {additional_holds_needed}")
            print(f"  Additional segments needed: {additional_segments_needed}")
        
        extended_audio = original_audio[:intro_end_sample].copy()
        
        for i in range(additional_segments_needed + 1):
            segment = original_audio[repeat_start_sample:repeat_end_sample].copy()
            if len(extended_audio.shape) > 1 and len(segment.shape) > 1:
                extended_audio = np.vstack([extended_audio, segment])
            else:
                if len(extended_audio.shape) > 1:
                    segment_stereo = np.column_stack([segment, segment])
                    extended_audio = np.vstack([extended_audio, segment_stereo])
                elif len(segment.shape) > 1:
                    extended_mono = np.column_stack([extended_audio, extended_audio])
                    segment_mono = np.mean(segment, axis=1)
                    extended_audio = np.vstack([extended_mono, segment_mono])
                else:
                    extended_audio = np.concatenate([extended_audio, segment])
        
        if self.config['audio_extension']['preserve_ending']:
            ending_duration_sec = self.config['audio_extension']['ending_duration_sec']
            ending_samples = min(
                len(original_audio) - repeat_end_sample,
                int(ending_duration_sec * sample_rate)
            )
            ending_segment = original_audio[repeat_end_sample:repeat_end_sample+ending_samples].copy()
            
            if len(extended_audio.shape) > 1 and len(ending_segment.shape) > 1:
                extended_audio = np.vstack([extended_audio, ending_segment])
            else:
                if len(extended_audio.shape) > 1:
                    ending_stereo = np.column_stack([ending_segment, ending_segment])
                    extended_audio = np.vstack([extended_audio, ending_stereo])
                elif len(ending_segment.shape) > 1:
                    extended_mono = np.column_stack([extended_audio, extended_audio])
                    ending_mono = np.mean(ending_segment, axis=1)
                    extended_audio = np.vstack([extended_mono, ending_mono])
                else:
                    extended_audio = np.concatenate([extended_audio, ending_segment])
        
        extended_timestamps = timestamps_df.iloc[:intro_end_idx+1].copy()
        time_offset = 0
        sample_offset = 0
        milliseconds_offset = 0
        
        for i in range(additional_segments_needed + 1):
            segment_timestamps = timestamps_df.iloc[repeat_start_idx:repeat_end_idx+1].copy()
            
            if i > 0:
                segment_duration_sec = (timestamps_df.iloc[repeat_end_idx]['time_seconds'] - 
                                          timestamps_df.iloc[repeat_start_idx]['time_seconds'])
                segment_milliseconds = (timestamps_df.iloc[repeat_end_idx]['milliseconds'] - 
                                        timestamps_df.iloc[repeat_start_idx]['milliseconds'])
                time_offset += segment_duration_sec
                milliseconds_offset += segment_milliseconds
                sample_offset += segment_samples
                
                segment_timestamps['time_seconds'] += time_offset
                segment_timestamps['milliseconds'] += milliseconds_offset
                segment_timestamps['sample_index'] += sample_offset
                
                for idx, row in segment_timestamps.iterrows():
                    minutes = int(row['time_seconds'] // 60)
                    seconds = row['time_seconds'] % 60
                    segment_timestamps.at[idx, 'timestamp'] = f"{minutes:02d}:{seconds:.1f}"
            
            extended_timestamps = pd.concat([extended_timestamps, segment_timestamps])
        
        if self.config['audio_extension']['preserve_ending']:
            remaining_timestamps = timestamps_df.iloc[repeat_end_idx+1:].copy()
            if not remaining_timestamps.empty:
                remaining_timestamps['time_seconds'] += time_offset
                remaining_timestamps['milliseconds'] += milliseconds_offset
                remaining_timestamps['sample_index'] += sample_offset
                for idx, row in remaining_timestamps.iterrows():
                    minutes = int(row['time_seconds'] // 60)
                    seconds = row['time_seconds'] % 60
                    remaining_timestamps.at[idx, 'timestamp'] = f"{minutes:02d}:{seconds:.1f}"
                extended_timestamps = pd.concat([extended_timestamps, remaining_timestamps])
        
        extended_timestamps = extended_timestamps.reset_index(drop=True)
        
        timestamp_str = datetime.now().strftime("%Y%m%d_%H%M%S")
        extended_audio_path = os.path.join(self.audio_output_dir, f"extended_boxbreathing_{timestamp_str}.wav")
        sf.write(extended_audio_path, extended_audio, sample_rate)
        
        extended_timestamps_path = os.path.join(self.experiment_log_dir, f"extended_boxbreathing_timestamps_{timestamp_str}.csv")
        extended_timestamps.to_csv(extended_timestamps_path, index=False)
        
        if self.config['debug_mode']:
            print("Extended audio created:")
            print(f"  Original length: {len(original_audio)/sample_rate:.2f} seconds")
            print(f"  Extended length: {len(extended_audio)/sample_rate:.2f} seconds")
            print(f"  Extended audio saved to: {extended_audio_path}")
            print(f"  Extended timestamps saved to: {extended_timestamps_path}")
        
        self.extended_audio = extended_audio
        self.extended_audio_path = extended_audio_path
        self.extended_timestamps_df = extended_timestamps
        
        return extended_audio_path, extended_timestamps
    
    def load_stimuli(self):
        """Load all stimulus files."""
        if self.looming_stimuli and self.tactile_stimulus is not None:
            return self.looming_stimuli, self.tactile_stimulus
        
        looming_stimuli = {}
        for stim_type, path in self.looming_stimuli_paths.items():
            try:
                data, sr = sf.read(path)
                if self.sample_rate is None:
                    self.sample_rate = sr
                elif sr != self.sample_rate:
                    print(f"Warning: Sample rate mismatch for {stim_type}: {sr} vs {self.sample_rate}")
                looming_stimuli[stim_type] = data
                if self.config['debug_mode']:
                    print(f"Loaded {stim_type} looming stimulus: {len(data)} samples")
            except Exception as e:
                print(f"Error loading {stim_type} stimulus: {e}")
        
        try:
            tactile_data, sr = sf.read(self.tactile_stimulus_path)
            if self.sample_rate is None:
                self.sample_rate = sr
            elif sr != self.sample_rate:
                print(f"Warning: Sample rate mismatch for tactile: {sr} vs {self.sample_rate}")
            self.tactile_stimulus = tactile_data
            if self.config['debug_mode']:
                print(f"Loaded tactile stimulus: {len(tactile_data)} samples")
        except Exception as e:
            print(f"Error loading tactile stimulus: {e}")
            self.tactile_stimulus = None
        
        self.looming_stimuli = looming_stimuli
        return looming_stimuli, self.tactile_stimulus
    
    def generate_counterbalanced_design(self, participant_id):
        """
        Generate a counterbalanced design for a participant.
        For each condition (inhale, exhale, baseline) and for each SOA,
        the combination is repeated a number of times (repetitions). Jitter and noise type
        are randomized for each trial.
        """
        soa_conditions = self.config['soa_conditions_ms']
        jitter_options = self.config['jitter_options_ms']
        stimulus_types = list(LOOMING_STIMULI.keys())
        
        all_trials = []
        
        # Generate trials for each condition (inhale, exhale, baseline)
        for condition, count in self.trial_counts.items():
            if condition == 'catch' or count == 0:
                continue
            
            # Create full combinations for stimulus type and SOA
            stim_soa_combinations = list(product(stimulus_types, soa_conditions))
            reps_needed = self.config['repetitions']  # repeat each combination a set number of times
            full_combinations = stim_soa_combinations * reps_needed
            
            # Shuffle to randomize order and then create trial entries
            random.shuffle(full_combinations)
            for stim_type, soa in full_combinations:
                all_trials.append({
                    'participant_id': participant_id,
                    'trial_number': len(all_trials) + 1,
                    'stimulus_type': stim_type,         # Noise type
                    'trial_type': condition,             # Condition from config (inhalation, exhalation, baseline)
                    'soa_value_ms': soa,                 # Base SOA value in ms
                    'jitter_ms': random.choice(jitter_options),  # Random jitter
                    'is_tactile': True
                })
        
        # Add catch trials separately (if needed)
        catch_count = self.trial_counts.get('catch', 0)
        if catch_count > 0:
            catch_stim_combinations = list(product(stimulus_types, soa_conditions))
            while len(catch_stim_combinations) < catch_count:
                catch_stim_combinations.extend(catch_stim_combinations)
            catch_combinations = random.sample(catch_stim_combinations, catch_count)
            
            for stim_type, soa in catch_combinations:
                all_trials.append({
                    'participant_id': participant_id,
                    'trial_number': len(all_trials) + 1,
                    'stimulus_type': stim_type,
                    'trial_type': 'catch',
                    'soa_value_ms': soa,
                    'jitter_ms': random.choice(jitter_options),
                    'is_tactile': False
                })
        
        design_df = pd.DataFrame(all_trials)
        design_df = design_df.sample(frac=1, random_state=participant_id*100).reset_index(drop=True)
        design_df['trial_number'] = range(1, len(design_df) + 1)
        return design_df
    
    def assign_trials_to_breath_holds(self, design_df, timestamps_df):
        """Assign trials to specific breath holds while maintaining balance."""
        intro_end = self.config['audio_extension']['intro_segment_end']
        intro_end_match = timestamps_df[timestamps_df['timestamp'] == intro_end]
        intro_end_idx = intro_end_match.index[0] if not intro_end_match.empty else int(len(timestamps_df) * 0.1)
        if intro_end_match.empty:
            print(f"Warning: Using fallback intro_end_idx = {intro_end_idx}")
        
        usable_timestamps = timestamps_df.iloc[intro_end_idx+1:].copy()
        inhale_timestamps = usable_timestamps[usable_timestamps['breathphase'] == 'inhale']
        exhale_timestamps = usable_timestamps[usable_timestamps['breathphase'] == 'exhale']
        
        if self.config['debug_mode']:
            print("Assigning trials to breath holds:")
            print(f"  Available inhale holds: {len(inhale_timestamps)}")
            print(f"  Available exhale holds: {len(exhale_timestamps)}")
        
        inhale_indices = inhale_timestamps.index.tolist()
        exhale_indices = exhale_timestamps.index.tolist()
        trial_assignments = {}
        
        for condition in set(design_df['trial_type']):
            condition_trials = design_df[design_df['trial_type'] == condition]
            condition_count = len(condition_trials)
            inhale_count = condition_count // 2
            exhale_count = condition_count - inhale_count
            
            if self.config['debug_mode']:
                print(f"  Condition '{condition}': {condition_count} trials - {inhale_count} inhale, {exhale_count} exhale")
            
            if inhale_count > len(inhale_indices) or exhale_count > len(exhale_indices):
                print(f"Warning: Not enough breath holds for condition '{condition}'")
                print(f"  Need {inhale_count} inhale holds, have {len(inhale_indices)}")
                print(f"  Need {exhale_count} exhale holds, have {len(exhale_indices)}")
                inhale_count = min(inhale_count, len(inhale_indices))
                exhale_count = min(exhale_count, len(exhale_indices))
            
            for idx, trial_idx in zip(inhale_indices[:inhale_count], condition_trials.index[:inhale_count]):
                timestamp_row = timestamps_df.loc[idx]
                trial_assignments[trial_idx] = {
                    'timestamp_idx': idx,
                    'breathphase': 'inhale',
                    'timestamp': timestamp_row['timestamp'],
                    'milliseconds': timestamp_row['milliseconds'],
                    'sample_index': timestamp_row['sample_index'],
                    'time_seconds': timestamp_row['time_seconds']
                }
            
            for idx, trial_idx in zip(exhale_indices[:exhale_count], condition_trials.index[inhale_count:inhale_count+exhale_count]):
                timestamp_row = timestamps_df.loc[idx]
                trial_assignments[trial_idx] = {
                    'timestamp_idx': idx,
                    'breathphase': 'exhale',
                    'timestamp': timestamp_row['timestamp'],
                    'milliseconds': timestamp_row['milliseconds'],
                    'sample_index': timestamp_row['sample_index'],
                    'time_seconds': timestamp_row['time_seconds']
                }
            
            inhale_indices = inhale_indices[inhale_count:]
            exhale_indices = exhale_indices[exhale_count:]
        
        design_df['breathphase'] = design_df.index.map(lambda idx: trial_assignments.get(idx, {}).get('breathphase'))
        design_df['timestamp'] = design_df.index.map(lambda idx: trial_assignments.get(idx, {}).get('timestamp'))
        design_df['milliseconds'] = design_df.index.map(lambda idx: trial_assignments.get(idx, {}).get('milliseconds'))
        design_df['sample_index'] = design_df.index.map(lambda idx: trial_assignments.get(idx, {}).get('sample_index'))
        design_df['time_seconds'] = design_df.index.map(lambda idx: trial_assignments.get(idx, {}).get('time_seconds'))
        
        unassigned = design_df[design_df['breathphase'].isna()]
        if not unassigned.empty:
            print(f"Warning: {len(unassigned)} trials could not be assigned to breath holds")
            available_timestamps = timestamps_df.loc[
                ~timestamps_df.index.isin([a['timestamp_idx'] for a in trial_assignments.values() if 'timestamp_idx' in a])
            ]
            if not available_timestamps.empty:
                print(f"  Using fallback assignment with {len(available_timestamps)} available timestamps")
                for idx in unassigned.index:
                    if available_timestamps.empty:
                        break
                    ts_idx = random.choice(available_timestamps.index)
                    timestamp_row = available_timestamps.loc[ts_idx]
                    design_df.at[idx, 'breathphase'] = timestamp_row.get('breathphase', 'unknown')
                    design_df.at[idx, 'timestamp'] = timestamp_row['timestamp']
                    design_df.at[idx, 'milliseconds'] = timestamp_row['milliseconds']
                    design_df.at[idx, 'sample_index'] = timestamp_row['sample_index']
                    design_df.at[idx, 'time_seconds'] = timestamp_row['time_seconds']
                    available_timestamps = available_timestamps.drop(ts_idx)
        
        design_df['jittered_ms'] = design_df.apply(
            lambda row: row['milliseconds'] + row['jitter_ms'] if pd.notnull(row['milliseconds']) else None,
            axis=1
        )
        
        design_df['soa_ms'] = design_df.apply(
            lambda row: row['jittered_ms'] + row['soa_value_ms'] if pd.notnull(row['jittered_ms']) and row['trial_type'] != 'catch' else None, 
            axis=1
        )
        
        # Format additional timestamp columns (for logging purposes)
        design_df['timestamp_original'] = design_df.apply(
            lambda row: f"{int(row['milliseconds'] // 60000):02}:{(row['milliseconds'] % 60000) / 1000:.1f}" if pd.notnull(row['milliseconds']) else None,
            axis=1
        )
        
        design_df['timestamp_after_jitter'] = design_df.apply(
            lambda row: f"{int(row['jittered_ms'] // 60000):02}:{(row['jittered_ms'] % 60000) / 1000:.1f}" if pd.notnull(row['jittered_ms']) else None,
            axis=1
        )
        
        design_df['timestamp_with_soa'] = design_df.apply(
            lambda row: f"{int(row['soa_ms'] // 60000):02}:{(row['soa_ms'] % 60000) / 1000:.1f}" if pd.notnull(row['soa_ms']) else None,
            axis=1
        )
        
        sample_rate = self.sample_rate if self.sample_rate else 48000
        design_df['base_ms'] = design_df['milliseconds']
        design_df['jittered_sample'] = design_df.apply(
            lambda row: int(row['jittered_ms'] / 1000 * sample_rate) if pd.notnull(row['jittered_ms']) else None,
            axis=1
        )
        design_df['soa_sample'] = design_df.apply(
            lambda row: int(row['soa_ms'] / 1000 * sample_rate) if pd.notnull(row['soa_ms']) else None,
            axis=1
        )
        
        return design_df
    
    def generate_participant_audio(self, participant_id, design_df, base_audio_path):
        """Generate participant-specific audio files."""
        try:
            base_audio, sample_rate = sf.read(base_audio_path)
            if self.sample_rate is None:
                self.sample_rate = sample_rate
        except Exception as e:
            print(f"Error loading base audio: {e}")
            return None, None, None
        
        looming_stimuli, tactile_stimulus = self.load_stimuli()
        looming_audio = base_audio.copy()
        if len(base_audio.shape) > 1:
            tactile_audio = np.zeros((len(base_audio), base_audio.shape[1]), dtype=base_audio.dtype)
        else:
            tactile_audio = np.zeros(len(base_audio), dtype=base_audio.dtype)
        
        if tactile_stimulus is not None:
            if len(tactile_stimulus.shape) == 1 and len(tactile_audio.shape) > 1:
                tactile_stereo = np.column_stack((tactile_stimulus, tactile_stimulus))
            elif len(tactile_stimulus.shape) > 1 and len(tactile_audio.shape) == 1:
                tactile_stereo = np.mean(tactile_stimulus, axis=1)
            else:
                tactile_stereo = tactile_stimulus
        else:
            print("Warning: No tactile stimulus available")
            tactile_stereo = None
        
        injection_log = []
        
        for _, row in design_df.iterrows():
            if pd.isnull(row['jittered_sample']):
                continue
            looming_pos = int(row['jittered_sample'])
            tactile_pos = int(row['soa_sample']) if row['trial_type'] != 'catch' and pd.notnull(row['soa_sample']) else None
            
            stim_type = row['stimulus_type']
            if stim_type in looming_stimuli:
                looming_data = looming_stimuli[stim_type]
                if looming_pos + len(looming_data) <= len(looming_audio):
                    if len(looming_data.shape) == 2 and len(looming_audio.shape) == 2:
                        looming_audio[looming_pos:looming_pos+len(looming_data)] += looming_data
                    elif len(looming_data.shape) == 1 and len(looming_audio.shape) == 2:
                        for ch in range(looming_audio.shape[1]):
                            looming_audio[looming_pos:looming_pos+len(looming_data), ch] += looming_data
                    elif len(looming_data.shape) == 2 and len(looming_audio.shape) == 1:
                        looming_audio[looming_pos:looming_pos+len(looming_data)] += np.mean(looming_data, axis=1)
                    else:
                        looming_audio[looming_pos:looming_pos+len(looming_data)] += looming_data
                    
                    injection_log.append({
                        'trial_number': row['trial_number'],
                        'trial_type': row['trial_type'],
                        'stimulus_type': stim_type,
                        'breathphase': row.get('breathphase', 'unknown'),
                        'looming_position': looming_pos,
                        'looming_time_ms': row['jittered_ms']
                    })
            
            if tactile_pos is not None and tactile_stereo is not None:
                if tactile_pos + len(tactile_stereo) <= len(tactile_audio):
                    if len(tactile_stereo.shape) == 2 and len(tactile_audio.shape) == 2:
                        tactile_audio[tactile_pos:tactile_pos+len(tactile_stereo)] += tactile_stereo
                    elif len(tactile_stereo.shape) == 1 and len(tactile_audio.shape) == 2:
                        for ch in range(tactile_audio.shape[1]):
                            tactile_audio[tactile_pos:tactile_pos+len(tactile_stereo), ch] += tactile_stereo
                    elif len(tactile_stereo.shape) == 2 and len(tactile_audio.shape) == 1:
                        tactile_audio[tactile_pos:tactile_pos+len(tactile_stereo)] += np.mean(tactile_stereo, axis=1)
                    else:
                        tactile_audio[tactile_pos:tactile_pos+len(tactile_stereo)] += tactile_stereo
                    
                    for log_entry in injection_log:
                        if log_entry['trial_number'] == row['trial_number']:
                            log_entry['tactile_position'] = tactile_pos
                            log_entry['tactile_time_ms'] = row['soa_ms']
                            log_entry['soa_ms'] = row['soa_value_ms']
                            break
        
        looming_filename = f"participant_{participant_id}_design_looming.wav"
        tactile_filename = f"participant_{participant_id}_design_tactile.wav"
        log_filename = f"participant_{participant_id}_design_stimuli_log.csv"
        
        looming_path = os.path.join(self.audio_output_dir, looming_filename)
        tactile_path = os.path.join(self.audio_output_dir, tactile_filename)
        log_path = os.path.join(self.audio_output_dir, log_filename)
        
        try:
            sf.write(looming_path, looming_audio, sample_rate)
            sf.write(tactile_path, tactile_audio, sample_rate)
            injection_df = pd.DataFrame(injection_log)
            injection_df.to_csv(log_path, index=False)
            
            if self.config['debug_mode']:
                print(f"Generated audio files for participant {participant_id}:")
                print(f"  Looming audio: {looming_path}")
                print(f"  Tactile audio: {tactile_path}")
                print(f"  Injection log: {log_path}")
                print(f"  Injected {len(injection_log)} stimuli")
            
            return looming_path, tactile_path, log_path
        except Exception as e:
            print(f"Error saving audio files: {e}")
            return None, None, None
    
    def verify_audio_synchronization(self, participant_id, design_df, injection_log_path):
        """Verify the synchronization between generated audio files and design."""
        results = {
            'success': False,
            'error': None,
            'warnings': []
        }
        
        try:
            injection_df = pd.read_csv(injection_log_path)
            design_count = len(design_df)
            injection_count = len(injection_df)
            
            if design_count != injection_count:
                warning = f"Design has {design_count} trials but injection log has {injection_count} trials"
                results['warnings'].append(warning)
            
            sample_size = min(5, len(injection_df))
            sample_trials = injection_df.sample(n=sample_size) if len(injection_df) > 5 else injection_df
            
            for _, row in sample_trials.iterrows():
                trial_number = row['trial_number']
                design_row = design_df[design_df['trial_number'] == trial_number]
                if len(design_row) == 0:
                    results['warnings'].append(f"Trial {trial_number} not found in design")
                    continue
                design_row = design_row.iloc[0]
                if 'looming_time_ms' in row and 'jittered_ms' in design_row:
                    looming_diff = abs(row['looming_time_ms'] - design_row['jittered_ms'])
                    if looming_diff > 1:
                        results['warnings'].append(f"Trial {trial_number}: Looming timing mismatch: {looming_diff} ms")
                if design_row['trial_type'] != 'catch' and 'tactile_time_ms' in row and 'soa_ms' in design_row:
                    tactile_diff = abs(row['tactile_time_ms'] - design_row['soa_ms'])
                    if tactile_diff > 1:
                        results['warnings'].append(f"Trial {trial_number}: Tactile timing mismatch: {tactile_diff} ms")
            
            results['success'] = True
            
            if self.config['debug_mode']:
                print(f"Verification complete for participant {participant_id}")
                if results['warnings']:
                    print(f"  Warnings: {len(results['warnings'])}")
                    for warning in results['warnings'][:3]:
                        print(f"    - {warning}")
                    if len(results['warnings']) > 3:
                        print(f"    - (and {len(results['warnings']) - 3} more warnings)")
                else:
                    print("  No warnings - all timings match expected values")
            
            return results
        except Exception as e:
            results['error'] = f"Error during verification: {str(e)}"
            results['success'] = False
            if self.config['debug_mode']:
                print(results['error'])
            return results
    
    def process_participant(self, participant_id):
        """Process a single participant from design generation to audio creation."""
        results = {
            'participant_id': participant_id,
            'status': 'error',
            'files': {},
            'errors': []
        }
        
        try:
            print(f"\n=== Processing participant {participant_id} ===")
            
            if self.extended_audio_path is None:
                base_audio_path, timestamps_df = self.extend_breathing_audio()
            else:
                base_audio_path = self.extended_audio_path
                timestamps_df = self.extended_timestamps_df
            
            design_df = self.generate_counterbalanced_design(participant_id)
            design_df = self.assign_trials_to_breath_holds(design_df, timestamps_df)
            
            # --- Finalize Participant Design Log ---
            # Create a new column 'trial_condition' mapping the condition keys to more intuitive labels.
            mapping = {"inhalation": "inhale", "exhalation": "exhale", "baseline": "baseline", "catch": "catch"}
            design_df['trial_condition'] = design_df['trial_type'].map(mapping)
            
            # Reorder and select columns for the participant log CSV.
            final_columns = [
                "participant_id", "trial_number", "trial_condition", "breathphase",
                "timestamp", "milliseconds", "jitter_ms", "soa_value_ms", "soa_ms",
                "stimulus_type", "jittered_ms", "jittered_sample", "soa_sample"
            ]
            design_df = design_df[final_columns]
            
            design_path = os.path.join(self.experiment_log_dir, f"participant_{participant_id}_design.csv")
            design_df.to_csv(design_path, index=False)
            results['files']['design'] = design_path
            
            looming_path, tactile_path, log_path = self.generate_participant_audio(participant_id, design_df, base_audio_path)
            
            if looming_path and tactile_path and log_path:
                results['files']['looming_audio'] = looming_path
                results['files']['tactile_audio'] = tactile_path
                results['files']['stimulus_log'] = log_path
            else:
                error_msg = "Failed to generate audio files"
                results['errors'].append(error_msg)
                print(f"Error: {error_msg}")
            
            if log_path:
                verification_results = self.verify_audio_synchronization(participant_id, design_df, log_path)
                if verification_results.get('warnings'):
                    print(f"Warning: {len(verification_results['warnings'])} synchronization issues detected")
            
            if not results['errors']:
                results['status'] = 'success'
                print(f"=== Successfully processed participant {participant_id} ===\n")
            else:
                print(f"=== Completed processing for participant {participant_id} with errors ===\n")
            
            return results
        except Exception as e:
            error_msg = f"Error processing participant {participant_id}: {str(e)}"
            results['errors'].append(error_msg)
            print(error_msg)
            traceback.print_exc()
            return results
    
    def run(self, participant_ids=None, num_participants=None):
        """Run the full audio generation process for multiple participants."""
        if participant_ids is None and num_participants is None:
            num_participants = self.config.get('num_participants', 5)
        
        if participant_ids is None:
            participant_ids = list(range(self.config.get('num_participants', 5)))
        
        results = {}
        print("=== Starting PPSAudioGenerator ===")
        print(f"Repetitions per condition: {self.config['repetitions']}")
        print(f"Processing {len(participant_ids)} participants")
        print("Output directories:")
        print(f"  Design files: {self.experiment_log_dir}")
        print(f"  Audio files: {self.audio_output_dir}")
        total_trials = sum(self.trial_counts.values())
        print(f"Total trials per participant: {total_trials}")
        
        for p_id in participant_ids:
            results[p_id] = self.process_participant(p_id)
        
        success_count = sum(1 for r in results.values() if r['status'] == 'success')
        print("\n=== Processing Summary ===")
        print(f"Successfully processed {success_count} of {len(participant_ids)} participants")
        
        timestamp_str = datetime.now().strftime("%Y%m%d_%H%M%S")
        report_path = os.path.join(self.experiment_log_dir, f"generation_report_{timestamp_str}.json")
        
        try:
            with open(report_path, 'w') as f:
                report_data = {
                    'timestamp': timestamp_str,
                    'participants': {
                        p_id: {
                            'status': r['status'],
                            'files': {k: os.path.basename(v) for k, v in r.get('files', {}).items()}
                        } for p_id, r in results.items()
                    }
                }
                json.dump(report_data, f, indent=2)
            print(f"Saved generation report to {report_path}")
        except Exception as e:
            print(f"Error saving generation report: {e}")
        
        print("=== Completed audio generation ===")
        return results

# ======================================================================
# MAIN FUNCTION
# ======================================================================

def main():
    """Main function to run the PPS Audio Generator."""
    generator = PPSAudioGenerator(CONFIG)
    results = generator.run(num_participants=CONFIG.get('num_participants', 5))
    
    for p_id, data in results.items():
        status = data['status']
        if status == 'success':
            print(f"Participant {p_id}: Success")
            for file_type, file_path in data.get('files', {}).items():
                print(f"  {file_type}: {os.path.basename(file_path)}")
        else:
            print(f"Participant {p_id}: Error - {', '.join(data.get('errors', ['Unknown error']))}")

if __name__ == "__main__":
    main()
```

```python
import os
import numpy as np
import pandas as pd
import random
import json
from itertools import product
from datetime import datetime
import traceback

# ======================================================================
# CONFIGURATION PARAMETERS
# ======================================================================

CONFIG = {
    'repetitions': 4,            # how many times each condition × SOA is repeated
    'num_participants': 5,       # default number of participants
    'included_conditions': {
        'inhalation': True,
        'exhalation': True,
        'baseline': True,
    },
    'catch_trial_percentage': 10,
    'soa_conditions_ms': [190, 400, 700, 1000, 1500],
    'jitter_options_ms': [100, 200, 300, 400, 500],
    'debug_mode': True,
    # CSV file with all the breath-hold timestamps
    'timestamps_csv': r"C:\Users\cogpsy-vrlab\Documents\GitHub\BreathingSpace\Level0_StimulusFiles\BoxBreathing_timestamps.csv",
    'paths': {
        'experiment_log_dir': r"C:\Users\cogpsy-vrlab\Documents\GitHub\BreathingSpace\Level1_AudioGeneration\ExperimentLog",
    }
}

# Noise types for looming (these can be any labels you want)
LOOMING_STIMULI = ['blue', 'brown', 'pink', 'white']

# ======================================================================
# CLASS: PPSDesignGenerator
# ======================================================================

class PPSDesignGenerator:
    """Generates CSV design files for Peripersonal Space (PPS) experiments (no audio)."""
    
    def __init__(self, config=None):
        self.config = config or CONFIG
        
        # Prepare output directory
        self.experiment_log_dir = self.config['paths']['experiment_log_dir']
        os.makedirs(self.experiment_log_dir, exist_ok=True)
        
        # Calculate trial counts
        self.trial_counts = self._calculate_trial_counts()
        self.conditions = [cond for cond, inc in self.config['included_conditions'].items() if inc]
        
        # Internal property for the timestamps
        self.timestamps_df = None
        
        if self.config['debug_mode']:
            print("Initialized PPSDesignGenerator with:")
            print(f"  - Repetitions: {self.config['repetitions']}")
            print(f"  - Conditions: {self.conditions}")
            print(f"  - SOA values: {self.config['soa_conditions_ms']}")
            print(f"  - Total trials per participant: {sum(self.trial_counts.values())}")
            print(f"  - Output directory: {self.experiment_log_dir}")
    
    def _calculate_trial_counts(self):
        """Calculate how many trials for each condition (inhale/exhale/baseline) plus catch."""
        reps = self.config['repetitions']
        soa_count = len(self.config['soa_conditions_ms'])
        stimulus_types = LOOMING_STIMULI
        catch_percentage = self.config['catch_trial_percentage']
        
        trial_counts = {}
        for condition, include in self.config['included_conditions'].items():
            if include:
                # (# noise types) × (# SOA) × (reps)
                trial_counts[condition] = len(stimulus_types) * soa_count * reps
        
        regular_trials = sum(trial_counts.values())
        catch_trials = int(np.ceil(regular_trials * catch_percentage / 100))
        trial_counts['catch'] = catch_trials
        
        if self.config['debug_mode']:
            print("Trial count calculation:")
            for cond, count in trial_counts.items():
                print(f"  {cond}: {count} trials")
            print(f"  Total: {sum(trial_counts.values())} trials")
        
        return trial_counts
    
    def parse_timestamps(self):
        """Parse the box breathing timestamps CSV once, storing in self.timestamps_df."""
        if self.timestamps_df is not None:
            return self.timestamps_df
        
        csv_path = self.config['timestamps_csv']
        if not os.path.exists(csv_path):
            print(f"ERROR: Timestamps CSV not found at {csv_path}")
            self.timestamps_df = pd.DataFrame()
            return self.timestamps_df
        
        try:
            df = pd.read_csv(csv_path)
            # Ensure 'breathphase' column is present
            if 'breathphase' not in df.columns:
                # Alternate inhale/exhale
                df['breathphase'] = ['inhale' if i % 2 == 0 else 'exhale' for i in range(len(df))]
            # Add time_seconds for convenience
            if 'time_seconds' not in df.columns:
                df['time_seconds'] = df['milliseconds'] / 1000.0
            
            self.timestamps_df = df
            
            if self.config['debug_mode']:
                print(f"Parsed {len(df)} timestamps from {csv_path}")
            
            return df
        except Exception as e:
            print(f"Error parsing timestamps CSV: {e}")
            traceback.print_exc()
            self.timestamps_df = pd.DataFrame()
            return self.timestamps_df
    
    def generate_counterbalanced_design(self, participant_id):
        """
        Generate a DataFrame with all trials (including baseline, catch, etc.)
        with random jitter, SOA, noise type, etc.
        """
        soa_conditions = self.config['soa_conditions_ms']
        jitter_options = self.config['jitter_options_ms']
        stimulus_types = LOOMING_STIMULI
        
        all_trials = []
        
        # Build the main conditions (inhale, exhale, baseline)
        for condition, count in self.trial_counts.items():
            if condition == 'catch' or count == 0:
                continue
            
            # All combos of (noise type × SOA), repeated 'reps' times
            stim_soa_combinations = list(product(stimulus_types, soa_conditions))
            reps_needed = self.config['repetitions']
            
            full_combinations = stim_soa_combinations * reps_needed
            random.shuffle(full_combinations)
            
            for stim_type, soa in full_combinations:
                all_trials.append({
                    'participant_id': participant_id,
                    'trial_type': condition,  # inhalation, exhalation, baseline
                    'stimulus_type': stim_type,
                    'soa_value_ms': soa,
                    'jitter_ms': random.choice(jitter_options),
                    'is_tactile': True  # baseline & main trials have tactile
                })
        
        # Add catch trials (looming only, no tactile)
        catch_count = self.trial_counts.get('catch', 0)
        if catch_count > 0:
            catch_combos = list(product(stimulus_types, soa_conditions))
            while len(catch_combos) < catch_count:
                catch_combos.extend(catch_combos)
            catch_sample = random.sample(catch_combos, catch_count)
            
            for stim_type, soa in catch_sample:
                all_trials.append({
                    'participant_id': participant_id,
                    'trial_type': 'catch',
                    'stimulus_type': stim_type,
                    'soa_value_ms': soa,
                    'jitter_ms': random.choice(jitter_options),
                    'is_tactile': False
                })
        
        # Convert to DataFrame and randomize order
        design_df = pd.DataFrame(all_trials)
        design_df = design_df.sample(frac=1, random_state=participant_id*100).reset_index(drop=True)
        
        # Add a trial_number
        design_df['trial_number'] = range(1, len(design_df) + 1)
        
        return design_df
    
    def assign_breath_holds(self, design_df):
        """
        Assign each trial to an inhale or exhale timestamp from the CSV.
        Also compute derived timestamps in 'MM:SS.s' format for retention,
        looming, and tactile. This is the main place we do any time offset logic.
        """
        timestamps_df = self.parse_timestamps()
        if timestamps_df.empty:
            # If we have no timestamps, just return the design with no assignment
            design_df['breathphase'] = None
            design_df['milliseconds'] = None
            design_df['jittered_ms'] = None
            design_df['soa_ms'] = None
            design_df['timestamp_original'] = None
            design_df['timestamp_after_jitter'] = None
            design_df['timestamp_with_soa'] = None
            return design_df
        
        # We'll assume the first ~10% are "intro" and skip them for assignment
        skip_count = int(len(timestamps_df) * 0.1)
        usable = timestamps_df.iloc[skip_count:].copy()
        
        # Separate inhale vs exhale
        inhale_df = usable[usable['breathphase'] == 'inhale'].index.tolist()
        exhale_df = usable[usable['breathphase'] == 'exhale'].index.tolist()
        
        # For each condition, we attempt to split half the trials to inhale, half to exhale
        trial_assignments = {}
        
        # Group by condition
        for condition in design_df['trial_type'].unique():
            condition_mask = (design_df['trial_type'] == condition)
            condition_indices = design_df[condition_mask].index
            condition_count = len(condition_indices)
            
            # naive approach: half to inhale, half to exhale
            half_inhale = condition_count // 2
            half_exhale = condition_count - half_inhale
            
            # Assign from the front of each list
            used_inhales = inhale_df[:half_inhale]
            used_exhales = exhale_df[:half_exhale]
            
            # Actually map them onto the design indices
            for d_idx, t_idx in zip(condition_indices[:half_inhale], used_inhales):
                row_ts = timestamps_df.loc[t_idx]
                trial_assignments[d_idx] = row_ts
            
            for d_idx, t_idx in zip(condition_indices[half_inhale:half_inhale+half_exhale], used_exhales):
                row_ts = timestamps_df.loc[t_idx]
                trial_assignments[d_idx] = row_ts
            
            # Remove those from the "pool"
            inhale_df = inhale_df[half_inhale:]
            exhale_df = exhale_df[half_exhale:]
        
        # Now fill in columns
        design_df['breathphase'] = design_df.index.map(
            lambda i: trial_assignments[i]['breathphase'] if i in trial_assignments else None
        )
        design_df['milliseconds'] = design_df.index.map(
            lambda i: trial_assignments[i]['milliseconds'] if i in trial_assignments else None
        )
        
        # Jittered_ms = base + jitter
        design_df['jittered_ms'] = design_df.apply(
            lambda row: row['milliseconds'] + row['jitter_ms'] if pd.notnull(row['milliseconds']) else None,
            axis=1
        )
        
        # If not a catch trial, tactile is offset from looming by the SOA
        design_df['soa_ms'] = design_df.apply(
            lambda row: row['jittered_ms'] + row['soa_value_ms']
            if (pd.notnull(row['jittered_ms']) and row['trial_type'] != 'catch') else None,
            axis=1
        )
        
        # Convert ms → "MM:SS.s"
        def ms_to_timestamp(ms):
            if pd.isnull(ms):
                return None
            m = int(ms // 60000)
            s = (ms % 60000) / 1000.0
            return f"{m:02}:{s:.1f}"
        
        design_df['timestamp_original'] = design_df['milliseconds'].apply(ms_to_timestamp)
        design_df['timestamp_after_jitter'] = design_df['jittered_ms'].apply(ms_to_timestamp)
        design_df['timestamp_with_soa'] = design_df['soa_ms'].apply(ms_to_timestamp)
        
        return design_df
    
    def finalize_design_csv(self, design_df):
        """
        1) Rename columns to final naming:
           - stimulus_type -> looming_stimulus_type
           - timestamp_original -> retentionphase_timestamp
           - timestamp_after_jitter -> looming_stimulus_timestamp
           - timestamp_with_soa -> tactile_stimulus_timestamp
        2) For baseline trials: set looming_stimulus_type='none', looming_stimulus_timestamp=NaN
        3) For catch trials: set tactile_stimulus_timestamp=NaN
        4) Return the final DataFrame (ready to save).
        """
        # Rename stimulus_type -> looming_stimulus_type
        design_df.rename(columns={'stimulus_type': 'looming_stimulus_type'},
                         inplace=True)
        
        # baseline => no looming
        baseline_mask = (design_df['trial_type'] == 'baseline')
        design_df.loc[baseline_mask, 'looming_stimulus_type'] = 'none'
        
        # Rename these columns
        design_df.rename(columns={
            'timestamp_original':      'retentionphase_timestamp',
            'timestamp_after_jitter':  'looming_stimulus_timestamp',
            'timestamp_with_soa':      'tactile_stimulus_timestamp'
        }, inplace=True)
        
        # baseline => looming_stimulus_timestamp=NaN
        design_df.loc[baseline_mask, 'looming_stimulus_timestamp'] = np.nan
        
        # catch => no tactile => tactile_stimulus_timestamp=NaN
        catch_mask = (design_df['trial_type'] == 'catch')
        design_df.loc[catch_mask, 'tactile_stimulus_timestamp'] = np.nan
        
        # Optionally reorder columns
        final_cols = [
            'participant_id',
            'trial_number',
            'trial_type',             # inhalation, exhalation, baseline, catch
            'breathphase',            # inhale or exhale
            'retentionphase_timestamp',
            'looming_stimulus_timestamp',
            'tactile_stimulus_timestamp',
            'looming_stimulus_type',
            'soa_value_ms',
            'jitter_ms',
            'is_tactile'
        ]
        # Keep only columns that actually exist in the DF
        final_cols = [c for c in final_cols if c in design_df.columns]
        design_df = design_df[final_cols]
        
        return design_df
    
    def process_participant(self, participant_id):
        """
        Generate the design, assign breath holds, finalize columns,
        and save as a CSV for one participant.
        """
        try:
            if self.config['debug_mode']:
                print(f"\n=== Processing participant {participant_id} ===")
            
            # 1) Generate base design (counterbalanced combos)
            design_df = self.generate_counterbalanced_design(participant_id)
            
            # 2) Assign each trial to a breath hold (inhale or exhale)
            design_df = self.assign_breath_holds(design_df)
            
            # 3) Finalize columns (rename, set baseline/catch to NA, etc.)
            design_df = self.finalize_design_csv(design_df)
            
            # 4) Save CSV
            out_path = os.path.join(self.experiment_log_dir,
                                    f"participant_{participant_id}_design.csv")
            design_df.to_csv(out_path, index=False)
            
            if self.config['debug_mode']:
                print(f"Saved participant {participant_id} design to {out_path}")
                print(design_df.head())
            
            return True, out_path
        except Exception as e:
            print(f"Error processing participant {participant_id}: {e}")
            traceback.print_exc()
            return False, None
    
    def run(self, participant_ids=None, num_participants=None):
        """
        Generate the design CSVs for multiple participants.
        """
        if participant_ids is None and num_participants is None:
            num_participants = self.config.get('num_participants', 5)
        
        if participant_ids is None:
            participant_ids = list(range(num_participants))
        
        if self.config['debug_mode']:
            print("=== Starting PPSDesignGenerator ===")
            print(f"Repetitions: {self.config['repetitions']}")
            print(f"Participants: {participant_ids}")
        
        results = {}
        for pid in participant_ids:
            success, path = self.process_participant(pid)
            results[pid] = {
                'success': success,
                'csv_path': path
            }
        
        return results

# ======================================================================
# MAIN
# ======================================================================

def main():
    generator = PPSDesignGenerator(CONFIG)
    results = generator.run()  # by default, uses num_participants=5
    
    # Print summary
    print("\n=== Generation Summary ===")
    for pid, info in results.items():
        if info['success']:
            print(f"Participant {pid}: CSV generated at {info['csv_path']}")
        else:
            print(f"Participant {pid}: ERROR - no CSV generated.")

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import numpy as np
import os

# Path to your original box-breathing timestamps CSV
TIMESTAMPS_CSV = r"C:\Users\cogpsy-vrlab\Documents\GitHub\BreathingSpace\Level0_StimulusFiles\BoxBreathing_timestamps.csv"

# Example: let's say we need 60 total retention-phase slots
# (i.e. we have 60 trials needing inhale/exhale holds).
TOTAL_NEEDED_SLOTS = 60

# Segment boundaries for the portion we want to repeat
SEGMENT_START_TIMESTAMP = "02:08.4"  # 13th hold ends
SEGMENT_END_TIMESTAMP   = "04:56.4"  # 34th hold ends

# Output path for our final CSV
OUTPUT_CSV = "extended_boxbreathing_timestamps.csv"


def main():
    # 1) Read the original timestamps CSV
    df = pd.read_csv(TIMESTAMPS_CSV)
    
    # Ensure the DataFrame has the columns we need
    if not all(col in df.columns for col in ["timestamp", "milliseconds", "sample_index", "breathphase"]):
        print("ERROR: CSV missing one of the required columns (timestamp, milliseconds, sample_index, breathphase).")
        return
    
    # Sort by time if not already
    # (Sometimes CSV can be out of order)
    df = df.sort_values(by="milliseconds").reset_index(drop=True)
    
    # 2) Identify row indices for the segment boundaries
    seg_start_rows = df.index[df["timestamp"] == SEGMENT_START_TIMESTAMP].tolist()
    seg_end_rows   = df.index[df["timestamp"] == SEGMENT_END_TIMESTAMP].tolist()
    
    if not seg_start_rows or not seg_end_rows:
        print(f"ERROR: Could not find segment boundaries {SEGMENT_START_TIMESTAMP} or {SEGMENT_END_TIMESTAMP} in CSV.")
        return
    
    seg_start_idx = seg_start_rows[0]
    seg_end_idx   = seg_end_rows[0]
    
    if seg_end_idx <= seg_start_idx:
        print("ERROR: Segment end occurs before start in the CSV. Check your timestamps.")
        return
    
    # 3) Split the original DataFrame into:
    #    (A) The intro portion (everything up to seg_start_idx)
    #    (B) The repeat portion (seg_start_idx to seg_end_idx inclusive)
    #    (C) The outro portion (everything after seg_end_idx) - optional if you want to keep it
    intro_df  = df.iloc[:seg_start_idx].copy()
    repeat_df = df.iloc[seg_start_idx:seg_end_idx+1].copy()
    outro_df  = df.iloc[seg_end_idx+1:].copy()  # you can keep or discard
    
    # The repeat portion has N slots (e.g. 22). We'll replicate that as needed.
    n_slots_in_repeat_block = len(repeat_df)
    
    # 4) Check how many total slots we already have (intro + repeat + outro)
    #    In many designs, we only count the slots from intro + repeat, ignoring the outro for “active” holds.
    #    But you can adapt as needed.
    existing_slots = len(intro_df) + len(repeat_df)
    
    if existing_slots >= TOTAL_NEEDED_SLOTS:
        # If we already have enough, no repetition needed
        print(f"No need to repeat; existing {existing_slots} slots >= {TOTAL_NEEDED_SLOTS} needed.")
        # Just set repetition=0 for everything
        intro_df["repetition"]  = 0
        repeat_df["repetition"] = 0
        # Combine them
        extended_df = pd.concat([intro_df, repeat_df, outro_df], ignore_index=True)
    else:
        # We need more slots
        slots_needed_after_intro = TOTAL_NEEDED_SLOTS - len(intro_df)
        # See how many additional repeat blocks we need
        # Each block adds n_slots_in_repeat_block new entries
        # We already have 1 block, so see how many more we need beyond the first
        blocks_needed = int(np.ceil(slots_needed_after_intro / n_slots_in_repeat_block)) - 1
        # e.g. if we need 60 total, have 12 intro, 22 repeat => 34 total. Need 26 more => 26/22=1.18 => 2 blocks total
        # But we already have 1, so we add 1 more beyond that, etc.
        
        # 5) Assign repetition=0 for intro and the first block
        intro_df["repetition"]  = 0
        repeat_df["repetition"] = 0
        
        # 6) Build additional blocks
        repeated_blocks = []
        
        # We track the offset in time_seconds, milliseconds, sample_index
        # by looking at the difference from the repeat block
        block_duration_sec = (repeat_df["milliseconds"].iloc[-1] - repeat_df["milliseconds"].iloc[0]) / 1000.0
        block_duration_samples = repeat_df["sample_index"].iloc[-1] - repeat_df["sample_index"].iloc[0]
        block_duration_ms = repeat_df["milliseconds"].iloc[-1] - repeat_df["milliseconds"].iloc[0]
        
        current_offset_sec   = 0.0
        current_offset_ms    = 0
        current_offset_samps = 0
        
        # Start offset from the end of the existing repeat block
        current_offset_sec   = block_duration_sec
        current_offset_ms    = block_duration_ms
        current_offset_samps = block_duration_samples
        
        for i in range(1, blocks_needed+1):
            block_copy = repeat_df.copy()
            # Now we offset the timestamps in block_copy
            # We'll shift them by the current_offset_ms, etc.
            
            # SHIFT: timestamps (human-readable)
            def shift_timestamp(ts):
                # parse "MM:SS.s"
                # A robust approach is to parse properly, but we can do approximate for demo:
                # We'll just parse out minutes, seconds
                if not isinstance(ts, str):
                    return ts
                parts = ts.split(":")
                if len(parts) != 2:
                    return ts
                mm = int(parts[0])
                ss = float(parts[1])
                total_sec = mm*60 + ss
                # shift by current_offset_sec
                shifted = total_sec + current_offset_sec
                # convert back to "MM:SS.s"
                new_mm = int(shifted // 60)
                new_ss = shifted % 60
                return f"{new_mm:02}:{new_ss:04.1f}"
            
            block_copy["timestamp"] = block_copy["timestamp"].apply(shift_timestamp)
            block_copy["milliseconds"] = block_copy["milliseconds"] + current_offset_ms
            block_copy["sample_index"] = block_copy["sample_index"] + current_offset_samps
            
            # Assign repetition = i
            block_copy["repetition"] = i
            
            repeated_blocks.append(block_copy)
            
            # Increase offset for next iteration
            current_offset_sec   += block_duration_sec
            current_offset_ms    += block_duration_ms
            current_offset_samps += block_duration_samples
        
        # Combine everything
        extended_df = pd.concat([intro_df, repeat_df] + repeated_blocks + [outro_df], ignore_index=True)
    
    # 7) Save the final extended DataFrame
    extended_df.to_csv(OUTPUT_CSV, index=False)
    print(f"Extended CSV saved to {OUTPUT_CSV}")
    print(f"Total entries: {len(extended_df)}")
    print(f"First few lines:\n{extended_df.head(8)}")
    print(f"Last few lines:\n{extended_df.tail(8)}")


if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import numpy as np

# Path to your original CSV
BOX_BREATHING_CSV = r"C:\Users\cogpsy-vrlab\Documents\GitHub\BreathingSpace\Level0_StimulusFiles\BoxBreathing_timestamps.csv"

# Boundaries for the segment we want to repeat
SEGMENT_START = "02:08.4"  # 13th hold ends
SEGMENT_END   = "04:56.4"  # 34th hold ends

def expand_retention_phases(total_needed):
    """
    Reads the original CSV of box-breathing timestamps, checks how many
    retention-phase rows it contains, and if needed, repeats the segment
    from SEGMENT_START to SEGMENT_END to supply enough rows.

    Returns a DataFrame with columns:
      timestamp, milliseconds, sample_index, breathphase, repetition
    where 'repetition' is 0 for the original rows, 1 for the first repeated block, etc.
    """
    # Load original
    df = pd.read_csv(BOX_BREATHING_CSV)
    # Sort by time (in case it's not sorted)
    df = df.sort_values(by="milliseconds").reset_index(drop=True)
    
    # Identify row indices for segment boundaries
    seg_start_idx_list = df.index[df["timestamp"] == SEGMENT_START].tolist()
    seg_end_idx_list   = df.index[df["timestamp"] == SEGMENT_END].tolist()
    if not seg_start_idx_list or not seg_end_idx_list:
        raise ValueError(f"Could not find start='{SEGMENT_START}' or end='{SEGMENT_END}' in CSV.")
    
    seg_start_idx = seg_start_idx_list[0]
    seg_end_idx   = seg_end_idx_list[0]
    if seg_end_idx <= seg_start_idx:
        raise ValueError("Segment end index is before start index. Check timestamps.")
    
    # Partition the CSV:
    # 1) intro portion: everything up to seg_start_idx
    # 2) repeat portion: from seg_start_idx to seg_end_idx
    # 3) outro portion: everything after seg_end_idx
    intro_df  = df.iloc[:seg_start_idx].copy()
    repeat_df = df.iloc[seg_start_idx:seg_end_idx+1].copy()
    outro_df  = df.iloc[seg_end_idx+1:].copy()
    
    # We'll label all original rows (intro + repeat + outro) with repetition=0 for clarity
    intro_df["repetition"]  = 0
    repeat_df["repetition"] = 0
    outro_df["repetition"]  = 0
    
    # The total # of rows in the original CSV is
    # len(df), but the "useful" retention phases might be just after some intro.
    # For simplicity, let's say all 3 parts are potentially used. 
    # (You can adapt to skip the outro if you prefer.)
    
    # Check how many total retention phases we have in the original
    original_count = len(df)
    if original_count >= total_needed:
        # Enough rows exist already
        print(f"Original CSV has {original_count} retention phases, enough for {total_needed} needed.")
        return df
    
    # Otherwise, we must replicate the repeat_df portion until we have enough
    # The repeat segment length
    seg_len = len(repeat_df)
    
    # How many we have if we keep the entire original
    current_count = original_count
    
    # We'll track offsets for time and samples based on the difference
    # between the start and end of the repeat block
    block_duration_ms = repeat_df["milliseconds"].iloc[-1] - repeat_df["milliseconds"].iloc[0]
    block_duration_samples = repeat_df["sample_index"].iloc[-1] - repeat_df["sample_index"].iloc[0]
    
    # For the human-readable "timestamp" column, we do a quick approximate shift
    # by the block duration in seconds (parsed from block_duration_ms).
    block_duration_sec = block_duration_ms / 1000.0
    
    # Function to shift the "MM:SS.s" by a certain # of seconds
    def shift_timestamp(ts, offset_sec):
        # parse "MM:SS.s"
        if not isinstance(ts, str):
            return ts
        parts = ts.split(":")
        if len(parts) != 2:
            return ts
        mm = int(parts[0])
        ss = float(parts[1])
        total_sec = mm * 60 + ss
        shifted_sec = total_sec + offset_sec
        new_mm = int(shifted_sec // 60)
        new_ss = shifted_sec % 60
        return f"{new_mm:02}:{new_ss:04.1f}"
    
    # Start offset after the last block
    # i.e. if the last row is at X ms, we start from there
    repeated_blocks = []
    offset_sec   = block_duration_sec
    offset_ms    = block_duration_ms
    offset_samps = block_duration_samples
    
    # We'll keep replicating until we have enough
    repetition_index = 1  # since 0 is for the original
    while current_count < total_needed:
        # copy the block
        block_copy = repeat_df.copy()
        
        # shift each row's timestamps
        block_copy["timestamp"] = block_copy["timestamp"].apply(lambda t: shift_timestamp(t, offset_sec))
        block_copy["milliseconds"] = block_copy["milliseconds"] + offset_ms
        block_copy["sample_index"] = block_copy["sample_index"] + offset_samps
        
        # set repetition
        block_copy["repetition"] = repetition_index
        
        repeated_blocks.append(block_copy)
        
        # increment counters
        repetition_index += 1
        current_count += seg_len
        
        # shift offsets for the next iteration
        offset_sec   += block_duration_sec
        offset_ms    += block_duration_ms
        offset_samps += block_duration_samples
    
    # Combine everything
    extended_df = pd.concat([intro_df, repeat_df] + repeated_blocks + [outro_df],
                            ignore_index=True)
    
    print(f"Needed {total_needed}, ended up with {len(extended_df)} retention-phase rows.")
    return extended_df


def main():
    # Suppose we have an experiment design that yields 80 total trials needed
    total_trials_needed = 80
    
    # Expand the CSV
    extended_timestamps_df = expand_retention_phases(total_trials_needed)
    
    # Save or use these timestamps for your experiment log
    out_csv = "ExtendedBoxBreathing_Timestamps.csv"
    extended_timestamps_df.to_csv(out_csv, index=False)
    print(f"Saved extended timestamps to: {out_csv}")
    
    # Next step: You would assign each of your 80 trials to one row in
    # extended_timestamps_df (matching inhale vs exhale, etc.).

if __name__ == "__main__":
    main()
```

```python
import os
import numpy as np
import pandas as pd
import random
import json
from itertools import product
from datetime import datetime
import traceback

# ======================================================================
# CONFIGURATION PARAMETERS
# ======================================================================

CONFIG = {
    'repetitions': 4,            # how many times each condition × SOA is repeated
    'num_participants': 5,       # default number of participants
    'included_conditions': {
        'inhalation': True,
        'exhalation': True,
        'baseline': True,
    },
    'catch_trial_percentage': 10,
    'soa_conditions_ms': [190, 400, 700, 1000, 1500],
    'jitter_options_ms': [100, 200, 300, 400, 500],
    'debug_mode': True,
    # CSV file with all the breath-hold timestamps
    'timestamps_csv': r"C:\Users\cogpsy-vrlab\Documents\GitHub\BreathingSpace\Level0_StimulusFiles\BoxBreathing_timestamps.csv",
    'paths': {
        'experiment_log_dir': r"C:\Users\cogpsy-vrlab\Documents\GitHub\BreathingSpace\Level1_AudioGeneration\ExperimentLog",
    }
}

# Noise types for looming (these can be any labels you want)
LOOMING_STIMULI = ['blue', 'brown', 'pink', 'white']

# The boundaries of the segment we will replicate if needed
SEGMENT_START = "02:08.4"  # 13th hold ends
SEGMENT_END   = "04:56.4"  # 34th hold ends

# ======================================================================
# CLASS: PPSDesignGenerator
# ======================================================================

class PPSDesignGenerator:
    """Generates CSV design files for Peripersonal Space (PPS) experiments (no audio)."""
    
    def __init__(self, config=None):
        self.config = config or CONFIG
        
        # Prepare output directory
        self.experiment_log_dir = self.config['paths']['experiment_log_dir']
        os.makedirs(self.experiment_log_dir, exist_ok=True)
        
        # Calculate trial counts
        self.trial_counts = self._calculate_trial_counts()
        self.conditions = [cond for cond, inc in self.config['included_conditions'].items() if inc]
        
        # Internal property for the timestamps
        self.timestamps_df = None
        
        if self.config['debug_mode']:
            print("Initialized PPSDesignGenerator with:")
            print(f"  - Repetitions: {self.config['repetitions']}")
            print(f"  - Conditions: {self.conditions}")
            print(f"  - SOA values: {self.config['soa_conditions_ms']}")
            print(f"  - Total trials per participant: {sum(self.trial_counts.values())}")
            print(f"  - Output directory: {self.experiment_log_dir}")
    
    def _calculate_trial_counts(self):
        """Calculate how many trials for each condition (inhale/exhale/baseline) plus catch."""
        reps = self.config['repetitions']
        soa_count = len(self.config['soa_conditions_ms'])
        stimulus_types = LOOMING_STIMULI
        catch_percentage = self.config['catch_trial_percentage']
        
        trial_counts = {}
        for condition, include in self.config['included_conditions'].items():
            if include:
                # (# noise types) × (# SOA) × (reps)
                trial_counts[condition] = len(stimulus_types) * soa_count * reps
        
        regular_trials = sum(trial_counts.values())
        catch_trials = int(np.ceil(regular_trials * catch_percentage / 100))
        trial_counts['catch'] = catch_trials
        
        if self.config['debug_mode']:
            print("Trial count calculation:")
            for cond, count in trial_counts.items():
                print(f"  {cond}: {count} trials")
            print(f"  Total: {sum(trial_counts.values())} trials")
        
        return trial_counts
    
    def _expand_timestamps_if_needed(self):
        """
        Expand self.timestamps_df if we don't have enough rows to cover all trials.
        We replicate the segment from SEGMENT_START to SEGMENT_END, offsetting times
        and labeling each repeated block with repetition=1,2,3,... The original rows
        get repetition=0.
        """
        if self.timestamps_df is None or self.timestamps_df.empty:
            return
        
        # 1) Figure out how many total trials we have
        #    (each trial needs exactly one breath hold).
        total_trials = sum(self.trial_counts.values())
        
        # 2) Check how many breath-hold rows are in the original CSV
        original_count = len(self.timestamps_df)
        if original_count >= total_trials:
            if self.config['debug_mode']:
                print(f"No expansion needed; CSV has {original_count} holds, need {total_trials}.")
            # Just set repetition=0 for all
            self.timestamps_df["repetition"] = 0
            return
        
        if self.config['debug_mode']:
            print(f"Expanding timestamps: have {original_count}, need {total_trials}.")
        
        # Sort the DataFrame by milliseconds to ensure order
        self.timestamps_df = self.timestamps_df.sort_values(by="milliseconds").reset_index(drop=True)
        
        # Identify the row indices for the segment we replicate
        seg_start_idx = self.timestamps_df.index[self.timestamps_df["timestamp"] == SEGMENT_START].tolist()
        seg_end_idx   = self.timestamps_df.index[self.timestamps_df["timestamp"] == SEGMENT_END].tolist()
        
        if not seg_start_idx or not seg_end_idx:
            raise ValueError(f"Could not find start='{SEGMENT_START}' or end='{SEGMENT_END}' in CSV.")
        
        start_i = seg_start_idx[0]
        end_i   = seg_end_idx[0]
        if end_i <= start_i:
            raise ValueError(f"Segment end index <= start index. Check your timestamps.")
        
        intro_df  = self.timestamps_df.iloc[:start_i].copy()
        repeat_df = self.timestamps_df.iloc[start_i:end_i+1].copy()
        outro_df  = self.timestamps_df.iloc[end_i+1:].copy()
        
        # Label original rows with repetition=0
        intro_df["repetition"]  = 0
        repeat_df["repetition"] = 0
        outro_df["repetition"]  = 0
        
        # Check how many we have if we keep the entire original CSV
        current_count = len(intro_df) + len(repeat_df) + len(outro_df)
        seg_len = len(repeat_df)
        
        if current_count >= total_trials:
            # We can just keep them as-is
            self.timestamps_df = pd.concat([intro_df, repeat_df, outro_df], ignore_index=True)
            return
        
        # Otherwise, replicate the repeat block until we have enough
        block_duration_ms = repeat_df["milliseconds"].iloc[-1] - repeat_df["milliseconds"].iloc[0]
        block_duration_samples = repeat_df["sample_index"].iloc[-1] - repeat_df["sample_index"].iloc[0]
        
        # For shifting the "timestamp" column (MM:SS.s)
        block_duration_sec = block_duration_ms / 1000.0
        
        def shift_timestamp(ts, offset_sec):
            if not isinstance(ts, str):
                return ts
            parts = ts.split(":")
            if len(parts) != 2:
                return ts
            mm = int(parts[0])
            ss = float(parts[1])
            total_s = mm*60 + ss
            new_s = total_s + offset_sec
            new_mm = int(new_s // 60)
            new_ss = new_s % 60
            return f"{new_mm:02}:{new_ss:04.1f}"
        
        repeated_blocks = []
        offset_sec   = block_duration_sec
        offset_ms    = block_duration_ms
        offset_samps = block_duration_samples
        
        repetition_index = 1  # original = 0, so first copy = 1
        
        while current_count < total_trials:
            block_copy = repeat_df.copy()
            
            # shift each row's timestamps
            block_copy["timestamp"] = block_copy["timestamp"].apply(lambda t: shift_timestamp(t, offset_sec))
            block_copy["milliseconds"] = block_copy["milliseconds"] + offset_ms
            block_copy["sample_index"] = block_copy["sample_index"] + offset_samps
            
            block_copy["repetition"] = repetition_index
            
            repeated_blocks.append(block_copy)
            
            # increment
            current_count += seg_len
            repetition_index += 1
            offset_sec   += block_duration_sec
            offset_ms    += block_duration_ms
            offset_samps += block_duration_samples
        
        # Combine everything
        self.timestamps_df = pd.concat([intro_df, repeat_df] + repeated_blocks + [outro_df], ignore_index=True)
        if self.config['debug_mode']:
            print(f"Final expanded CSV has {len(self.timestamps_df)} breath holds (needed {total_trials}).")
    
    def parse_timestamps(self):
        """Parse the box breathing timestamps CSV once, storing in self.timestamps_df, then expand if needed."""
        if self.timestamps_df is not None:
            return self.timestamps_df
        
        csv_path = self.config['timestamps_csv']
        if not os.path.exists(csv_path):
            print(f"ERROR: Timestamps CSV not found at {csv_path}")
            self.timestamps_df = pd.DataFrame()
            return self.timestamps_df
        
        try:
            df = pd.read_csv(csv_path)
            # Ensure 'breathphase' column is present
            if 'breathphase' not in df.columns:
                df['breathphase'] = ['inhale' if i % 2 == 0 else 'exhale' for i in range(len(df))]
            # Add time_seconds for convenience
            if 'time_seconds' not in df.columns:
                df['time_seconds'] = df['milliseconds'] / 1000.0
            
            # The original CSV, no repetition assigned yet
            df["repetition"] = 0
            
            self.timestamps_df = df
            
            if self.config['debug_mode']:
                print(f"Parsed {len(df)} timestamps from {csv_path}")
            
            # Now expand if needed
            self._expand_timestamps_if_needed()
            
            return self.timestamps_df
        except Exception as e:
            print(f"Error parsing timestamps CSV: {e}")
            traceback.print_exc()
            self.timestamps_df = pd.DataFrame()
            return self.timestamps_df
    
    def generate_counterbalanced_design(self, participant_id):
        """
        Generate a DataFrame with all trials (including baseline, catch, etc.)
        with random jitter, SOA, noise type, etc.
        """
        soa_conditions = self.config['soa_conditions_ms']
        jitter_options = self.config['jitter_options_ms']
        stimulus_types = LOOMING_STIMULI
        
        all_trials = []
        
        # Build the main conditions (inhale, exhale, baseline)
        for condition, count in self.trial_counts.items():
            if condition == 'catch' or count == 0:
                continue
            
            # All combos of (noise type × SOA), repeated 'reps' times
            stim_soa_combinations = list(product(stimulus_types, soa_conditions))
            reps_needed = self.config['repetitions']
            
            full_combinations = stim_soa_combinations * reps_needed
            random.shuffle(full_combinations)
            
            for stim_type, soa in full_combinations:
                all_trials.append({
                    'participant_id': participant_id,
                    'trial_type': condition,  # inhalation, exhalation, baseline
                    'stimulus_type': stim_type,
                    'soa_value_ms': soa,
                    'jitter_ms': random.choice(jitter_options),
                    'is_tactile': True  # baseline & main trials have tactile
                })
        
        # Add catch trials (looming only, no tactile)
        catch_count = self.trial_counts.get('catch', 0)
        if catch_count > 0:
            catch_combos = list(product(stimulus_types, soa_conditions))
            while len(catch_combos) < catch_count:
                catch_combos.extend(catch_combos)
            catch_sample = random.sample(catch_combos, catch_count)
            
            for stim_type, soa in catch_sample:
                all_trials.append({
                    'participant_id': participant_id,
                    'trial_type': 'catch',
                    'stimulus_type': stim_type,
                    'soa_value_ms': soa,
                    'jitter_ms': random.choice(jitter_options),
                    'is_tactile': False
                })
        
        # Convert to DataFrame and randomize order
        design_df = pd.DataFrame(all_trials)
        design_df = design_df.sample(frac=1, random_state=participant_id*100).reset_index(drop=True)
        
        # Add a trial_number
        design_df['trial_number'] = range(1, len(design_df) + 1)
        
        return design_df
    
    def assign_breath_holds(self, design_df):
        """
        Assign each trial to an inhale or exhale row in self.timestamps_df.
        Also compute derived timestamps in 'MM:SS.s' format for retention,
        looming, and tactile. 
        """
        timestamps_df = self.parse_timestamps()  # ensures expansion is done
        if timestamps_df.empty:
            # If we have no timestamps, just return the design with no assignment
            design_df['breathphase'] = None
            design_df['milliseconds'] = None
            design_df['jittered_ms'] = None
            design_df['soa_ms'] = None
            design_df['timestamp_original'] = None
            design_df['timestamp_after_jitter'] = None
            design_df['timestamp_with_soa'] = None
            return design_df
        
        # We'll assume the first ~10% are "intro" and skip them for assignment
        skip_count = int(len(timestamps_df) * 0.1)
        usable = timestamps_df.iloc[skip_count:].copy()
        
        # Separate inhale vs exhale
        inhale_df = usable[usable['breathphase'] == 'inhale'].index.tolist()
        exhale_df = usable[usable['breathphase'] == 'exhale'].index.tolist()
        
        # For each condition, we attempt to split half the trials to inhale, half to exhale
        trial_assignments = {}
        
        # Group by condition
        for condition in design_df['trial_type'].unique():
            condition_mask = (design_df['trial_type'] == condition)
            condition_indices = design_df[condition_mask].index
            condition_count = len(condition_indices)
            
            half_inhale = condition_count // 2
            half_exhale = condition_count - half_inhale
            
            used_inhales = inhale_df[:half_inhale]
            used_exhales = exhale_df[:half_exhale]
            
            # Actually map them onto the design indices
            for d_idx, t_idx in zip(condition_indices[:half_inhale], used_inhales):
                row_ts = timestamps_df.loc[t_idx]
                trial_assignments[d_idx] = row_ts
            
            for d_idx, t_idx in zip(condition_indices[half_inhale:half_inhale+half_exhale], used_exhales):
                row_ts = timestamps_df.loc[t_idx]
                trial_assignments[d_idx] = row_ts
            
            # Remove those from the "pool"
            inhale_df = inhale_df[half_inhale:]
            exhale_df = exhale_df[half_exhale:]
        
        # Now fill in columns
        design_df['breathphase'] = design_df.index.map(
            lambda i: trial_assignments[i]['breathphase'] if i in trial_assignments else None
        )
        design_df['milliseconds'] = design_df.index.map(
            lambda i: trial_assignments[i]['milliseconds'] if i in trial_assignments else None
        )
        
        # Jittered_ms = base + jitter
        design_df['jittered_ms'] = design_df.apply(
            lambda row: row['milliseconds'] + row['jitter_ms'] if pd.notnull(row['milliseconds']) else None,
            axis=1
        )
        
        # If not a catch trial, tactile is offset from looming by the SOA
        design_df['soa_ms'] = design_df.apply(
            lambda row: row['jittered_ms'] + row['soa_value_ms']
            if (pd.notnull(row['jittered_ms']) and row['trial_type'] != 'catch') else None,
            axis=1
        )
        
        # Convert ms → "MM:SS.s"
        def ms_to_timestamp(ms):
            if pd.isnull(ms):
                return None
            m = int(ms // 60000)
            s = (ms % 60000) / 1000.0
            return f"{m:02}:{s:.1f}"
        
        design_df['timestamp_original'] = design_df['milliseconds'].apply(ms_to_timestamp)
        design_df['timestamp_after_jitter'] = design_df['jittered_ms'].apply(ms_to_timestamp)
        design_df['timestamp_with_soa'] = design_df['soa_ms'].apply(ms_to_timestamp)
        
        return design_df
    
    def finalize_design_csv(self, design_df):
        """
        1) Rename columns to final naming:
           - stimulus_type -> looming_stimulus_type
           - timestamp_original -> retentionphase_timestamp
           - timestamp_after_jitter -> looming_stimulus_timestamp
           - timestamp_with_soa -> tactile_stimulus_timestamp
        2) For baseline trials: set looming_stimulus_type='none', looming_stimulus_timestamp=NaN
        3) For catch trials: set tactile_stimulus_timestamp=NaN
        4) Return the final DataFrame (ready to save).
        """
        # Rename stimulus_type -> looming_stimulus_type
        design_df.rename(columns={'stimulus_type': 'looming_stimulus_type'},
                         inplace=True)
        
        # baseline => no looming
        baseline_mask = (design_df['trial_type'] == 'baseline')
        design_df.loc[baseline_mask, 'looming_stimulus_type'] = 'none'
        
        # Rename these columns
        design_df.rename(columns={
            'timestamp_original':      'retentionphase_timestamp',
            'timestamp_after_jitter':  'looming_stimulus_timestamp',
            'timestamp_with_soa':      'tactile_stimulus_timestamp'
        }, inplace=True)
        
        # baseline => looming_stimulus_timestamp=NaN
        design_df.loc[baseline_mask, 'looming_stimulus_timestamp'] = np.nan
        
        # catch => no tactile => tactile_stimulus_timestamp=NaN
        catch_mask = (design_df['trial_type'] == 'catch')
        design_df.loc[catch_mask, 'tactile_stimulus_timestamp'] = np.nan
        
        # Optionally reorder columns
        final_cols = [
            'participant_id',
            'trial_number',
            'trial_type',             # inhalation, exhalation, baseline, catch
            'breathphase',            # inhale or exhale
            'retentionphase_timestamp',
            'looming_stimulus_timestamp',
            'tactile_stimulus_timestamp',
            'looming_stimulus_type',
            'soa_value_ms',
            'jitter_ms',
            'is_tactile'
        ]
        # Keep only columns that actually exist in the DF
        final_cols = [c for c in final_cols if c in design_df.columns]
        design_df = design_df[final_cols]
        
        return design_df
    
    def process_participant(self, participant_id):
        """
        Generate the design, assign breath holds, finalize columns,
        and save as a CSV for one participant.
        """
        try:
            if self.config['debug_mode']:
                print(f"\n=== Processing participant {participant_id} ===")
            
            # 1) Generate base design (counterbalanced combos)
            design_df = self.generate_counterbalanced_design(participant_id)
            
            # 2) Assign each trial to a breath hold (inhale or exhale)
            design_df = self.assign_breath_holds(design_df)
            
            # 3) Finalize columns (rename, set baseline/catch to NA, etc.)
            design_df = self.finalize_design_csv(design_df)
            
            # 4) Save CSV
            out_path = os.path.join(self.experiment_log_dir,
                                    f"participant_{participant_id}_design.csv")
            design_df.to_csv(out_path, index=False)
            
            if self.config['debug_mode']:
                print(f"Saved participant {participant_id} design to {out_path}")
                print(design_df.head())
            
            return True, out_path
        except Exception as e:
            print(f"Error processing participant {participant_id}: {e}")
            traceback.print_exc()
            return False, None
    
    def run(self, participant_ids=None, num_participants=None):
        """
        Generate the design CSVs for multiple participants.
        """
        if participant_ids is None and num_participants is None:
            num_participants = self.config.get('num_participants', 5)
        
        if participant_ids is None:
            participant_ids = list(range(num_participants))
        
        if self.config['debug_mode']:
            print("=== Starting PPSDesignGenerator ===")
            print(f"Repetitions: {self.config['repetitions']}")
            print(f"Participants: {participant_ids}")
        
        results = {}
        for pid in participant_ids:
            success, path = self.process_participant(pid)
            results[pid] = {
                'success': success,
                'csv_path': path
            }
        
        return results

# ======================================================================
# MAIN
# ======================================================================

def main():
    generator = PPSDesignGenerator(CONFIG)
    results = generator.run()  # by default, uses num_participants=5
    
    # Print summary
    print("\n=== Generation Summary ===")
    for pid, info in results.items():
        if info['success']:
            print(f"Participant {pid}: CSV generated at {info['csv_path']}")
        else:
            print(f"Participant {pid}: ERROR - no CSV generated.")

if __name__ == "__main__":
    main()
```
