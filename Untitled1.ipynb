```python
import sounddevice as sd
import soundfile as sf
import numpy as np
import time
import os

def list_audio_devices():
    """List all available audio devices with their IDs"""
    print("Available audio devices:")
    print("-" * 70)
    devices = sd.query_devices()
    for i, device in enumerate(devices):
        print(f"ID: {i}, Name: {device['name']}, Channels: {device['max_output_channels']}")
    print("-" * 70)
    return devices

def find_device_by_name(devices, name_fragment):
    """Find a device ID by a partial name match"""
    for i, device in enumerate(devices):
        if name_fragment.lower() in device['name'].lower():
            return i
    return None

def play_synchronized_stimuli(bluetooth_device_id, komplete_device_id, 
                             tactile_file, looming_file, 
                             latency_compensation=0.0):
    """
    Play tactile stimuli via Bluetooth and looming stimuli via Komplete interface
    
    Parameters:
    -----------
    bluetooth_device_id : int
        Device ID for the Bluetooth output (for tactile)
    komplete_device_id : int
        Device ID for the Komplete Audio interface (for looming)
    tactile_file : str
        Path to the audio file for tactile feedback
    looming_file : str
        Path to the audio file for looming visual stimulus
    latency_compensation : float
        Compensation time in seconds to account for Bluetooth latency
    """
    print(f"Loading tactile stimulus: {tactile_file}")
    print(f"Loading looming stimulus: {looming_file}")
    
    # Load audio files
    tactile_data, tactile_samplerate = sf.read(tactile_file)
    looming_data, looming_samplerate = sf.read(looming_file)
    
    # Ensure tactile is mono (for Bluetooth)
    if len(tactile_data.shape) > 1 and tactile_data.shape[1] > 1:
        tactile_data = tactile_data[:, 0]  # Take first channel
    
    # Ensure looming is stereo for Komplete output 1/2
    if len(looming_data.shape) == 1:
        # Convert mono to stereo by duplicating
        looming_data = np.column_stack((looming_data, looming_data))
    elif looming_data.shape[1] == 1:
        looming_data = np.column_stack((looming_data, looming_data))
    
    # Get device info
    devices = sd.query_devices()
    bluetooth_device = devices[bluetooth_device_id]
    komplete_device = devices[komplete_device_id]
    
    print(f"Bluetooth device: {bluetooth_device['name']}")
    print(f"Komplete device: {komplete_device['name']}")
    
    # Create output streams
    bluetooth_stream = sd.OutputStream(
        samplerate=tactile_samplerate,
        device=bluetooth_device_id,
        channels=1,
        dtype='float32'
    )
    
    komplete_stream = sd.OutputStream(
        samplerate=looming_samplerate,
        device=komplete_device_id,
        channels=2,  # Stereo output for channels 1/2
        dtype='float32'
    )
    
    # Prepare tactile data for output
    tactile_output = tactile_data.reshape(-1, 1).astype(np.float32)
    
    # Start streams
    bluetooth_stream.start()
    komplete_stream.start()
    
    print("Starting playback...")
    
    # Account for Bluetooth latency if specified
    if latency_compensation > 0:
        print(f"Compensating for Bluetooth latency: {latency_compensation} seconds")
        # Start looming stimulus after delay to compensate for Bluetooth latency
        time.sleep(latency_compensation)
        komplete_stream.write(looming_data)
        # Calculate remaining tactile playtime
        tactile_duration = len(tactile_data) / tactile_samplerate
        remaining_time = tactile_duration - latency_compensation
        if remaining_time > 0:
            time.sleep(remaining_time)
    else:
        # Start both simultaneously
        bluetooth_start = time.time()
        bluetooth_stream.write(tactile_output)
        komplete_stream.write(looming_data)
        bluetooth_end = time.time()
        print(f"Playback duration: {bluetooth_end - bluetooth_start:.2f} seconds")
    
    # Stop and close streams
    bluetooth_stream.stop()
    komplete_stream.stop()
    bluetooth_stream.close()
    komplete_stream.close()
    
    print("Playback complete")

def main():
    # List all available devices
    devices = list_audio_devices()
    
    # Try to automatically find the devices
    bluetooth_id = find_device_by_name(devices, "bluetooth")
    komplete_id = find_device_by_name(devices, "komplete")
    
    # Allow manual selection if automatic detection fails
    if bluetooth_id is None:
        print("Bluetooth device not automatically detected.")
        bluetooth_id = int(input("Enter ID for Bluetooth device: "))
    else:
        print(f"Detected Bluetooth device at ID {bluetooth_id}: {devices[bluetooth_id]['name']}")
        confirm = input("Use this device? (y/n): ")
        if confirm.lower() != 'y':
            bluetooth_id = int(input("Enter ID for Bluetooth device: "))
    
    if komplete_id is None:
        print("Komplete Audio interface not automatically detected.")
        komplete_id = int(input("Enter ID for Komplete Audio interface: "))
    else:
        print(f"Detected Komplete interface at ID {komplete_id}: {devices[komplete_id]['name']}")
        confirm = input("Use this device? (y/n): ")
        if confirm.lower() != 'y':
            komplete_id = int(input("Enter ID for Komplete Audio interface: "))
    
    # Ask for file paths or use defaults
    default_dir = os.path.dirname(os.path.abspath(__file__))
    
    tactile_default = os.path.join(default_dir, "tactile_stimulus.wav")
    looming_default = os.path.join(default_dir, "looming_stimulus.wav")
    
    tactile_file = input(f"Enter path to tactile stimulus file (default: {tactile_default}): ")
    if not tactile_file:
        tactile_file = tactile_default
    
    looming_file = input(f"Enter path to looming stimulus file (default: {looming_default}): ")
    if not looming_file:
        looming_file = looming_default
    
    # Ask for latency compensation
    latency_str = input("Enter Bluetooth latency compensation in milliseconds (default: 0): ")
    latency_compensation = float(latency_str)/1000 if latency_str else 0.0
    
    # Play stimuli
    play_synchronized_stimuli(bluetooth_id, komplete_id, tactile_file, looming_file, latency_compensation)
    
    # Ask if user wants to play again
    again = input("Play again? (y/n): ")
    while again.lower() == 'y':
        play_synchronized_stimuli(bluetooth_id, komplete_id, tactile_file, looming_file, latency_compensation)
        again = input("Play again? (y/n): ")

if __name__ == "__main__":
    main()
```

```python
import sounddevice as sd
import soundfile as sf
import numpy as np
import time
import os
import threading
from queue import Queue
import signal

# Global flag for terminating the program
terminate_program = False

def signal_handler(sig, frame):
    global terminate_program
    print("\nCaught Ctrl+C, stopping playback...")
    terminate_program = True

signal.signal(signal.SIGINT, signal_handler)

def list_audio_devices():
    """List all available audio devices with their IDs"""
    print("Available audio devices:")
    print("-" * 70)
    devices = sd.query_devices()
    for i, device in enumerate(devices):
        print(f"ID: {i}, Name: {device['name']}, Channels: {device['max_output_channels']}")
    print("-" * 70)
    return devices

def find_device_by_name(devices, name_fragment, exact_match=False):
    """Find a device ID by a partial or exact name match"""
    for i, device in enumerate(devices):
        if exact_match:
            if device['name'] == name_fragment:
                return i
        else:
            if name_fragment.lower() in device['name'].lower():
                return i
    return None

def find_woojer_device(devices):
    """Find Woojer device for tactile stimuli"""
    for i, device in enumerate(devices):
        if "woojer" in device['name'].lower() and device['max_output_channels'] > 0:
            return i
    return None

def find_looming_device(devices):
    """Find the exact 'Output 1/2 (Output 1/2)' device or an alternative Komplete device"""
    device_id = find_device_by_name(devices, "Output 1/2 (Output 1/2)", exact_match=True)
    
    # If exact device not found, try Komplete Audio devices
    if device_id is None:
        for i, device in enumerate(devices):
            if "komplete audio" in device['name'].lower() and "output 1/2" in device['name'].lower() and device['max_output_channels'] > 0:
                return i
    return device_id

def woojer_callback(outdata, frames, time, status):
    """Callback function for Woojer audio output"""
    global woojer_position, tactile_data
    
    if status:
        print(f"Woojer status: {status}")
    
    if woojer_position >= len(tactile_data):
        # End of data
        outdata.fill(0)
        return
    
    # Calculate how many frames we can still play
    remainder = len(tactile_data) - woojer_position
    frames_to_play = min(frames, remainder)
    
    if frames_to_play > 0:
        audio_chunk = tactile_data[woojer_position:woojer_position+frames_to_play]
        # Reshape for mono output
        outdata[:frames_to_play, 0] = audio_chunk
        # Fill the rest with zeros if we're at the end
        if frames_to_play < frames:
            outdata[frames_to_play:, 0] = 0
        woojer_position += frames_to_play
    else:
        outdata.fill(0)

def looming_callback(outdata, frames, time, status):
    """Callback function for looming audio output"""
    global looming_position, looming_data
    
    if status:
        print(f"Looming status: {status}")
    
    if looming_position >= len(looming_data):
        # End of data
        outdata.fill(0)
        return
    
    # Calculate how many frames we can still play
    remainder = len(looming_data) - looming_position
    frames_to_play = min(frames, remainder)
    
    if frames_to_play > 0:
        # Get stereo data
        audio_chunk = looming_data[looming_position:looming_position+frames_to_play]
        # Copy to output, handling both mono and stereo
        if len(audio_chunk.shape) == 1:  # Mono input
            outdata[:frames_to_play, 0] = audio_chunk
            outdata[:frames_to_play, 1] = audio_chunk
        else:  # Stereo input
            outdata[:frames_to_play] = audio_chunk
        # Fill the rest with zeros if we're at the end
        if frames_to_play < frames:
            outdata[frames_to_play:] = 0
        looming_position += frames_to_play
    else:
        outdata.fill(0)

def play_synchronized_stimuli(woojer_device_id, looming_device_id, 
                             tactile_file, looming_file, 
                             latency_compensation=0.0):
    """
    Play tactile stimuli via Woojer and looming stimuli via Output 1/2 interface
    using callback-based approach to avoid blocking API issues
    
    Parameters:
    -----------
    woojer_device_id : int
        Device ID for the Woojer output (for tactile)
    looming_device_id : int
        Device ID for the Output 1/2 interface (for looming)
    tactile_file : str
        Path to the audio file for tactile feedback
    looming_file : str
        Path to the audio file for looming visual stimulus
    latency_compensation : float
        Compensation time in seconds to account for Woojer latency
    """
    global tactile_data, looming_data, woojer_position, looming_position
    
    print(f"Loading tactile stimulus: {tactile_file}")
    print(f"Loading looming stimulus: {looming_file}")
    
    # Load audio files
    tactile_data, tactile_samplerate = sf.read(tactile_file)
    looming_data, looming_samplerate = sf.read(looming_file)
    
    # Ensure tactile is mono (for Woojer)
    if len(tactile_data.shape) > 1 and tactile_data.shape[1] > 1:
        tactile_data = tactile_data[:, 0]  # Take first channel
    
    # Ensure looming is stereo for Output 1/2
    if len(looming_data.shape) == 1:
        # Convert mono to stereo by duplicating
        looming_data = np.column_stack((looming_data, looming_data))
    elif looming_data.shape[1] == 1:
        looming_data = np.column_stack((looming_data, looming_data))
    
    # Get device info
    devices = sd.query_devices()
    woojer_device = devices[woojer_device_id]
    looming_device = devices[looming_device_id]
    
    print(f"Tactile device (Woojer): {woojer_device['name']}")
    print(f"Looming device: {looming_device['name']}")
    
    # Reset positions
    woojer_position = 0
    looming_position = 0
    
    # Calculate playback durations
    tactile_duration = len(tactile_data) / tactile_samplerate
    looming_duration = len(looming_data) / looming_samplerate
    
    print(f"Tactile duration: {tactile_duration:.2f} seconds")
    print(f"Looming duration: {looming_duration:.2f} seconds")
    
    # Create streams with callbacks
    woojer_stream = sd.OutputStream(
        samplerate=tactile_samplerate,
        device=woojer_device_id,
        channels=1,  # Mono output for tactile
        callback=woojer_callback,
        blocksize=1024
    )
    
    looming_stream = sd.OutputStream(
        samplerate=looming_samplerate,
        device=looming_device_id,
        channels=2,  # Stereo output for channels 1/2
        callback=looming_callback,
        blocksize=1024
    )
    
    # Start with latency compensation if needed
    print("\nStarting playback...")
    
    woojer_stream.start()
    
    # Apply latency compensation if specified
    if latency_compensation > 0:
        print(f"Compensating for Woojer latency: {latency_compensation} seconds")
        time.sleep(latency_compensation)
    
    looming_stream.start()
    
    # Wait for the longer of the two streams to finish
    max_duration = max(tactile_duration, looming_duration) + 0.5  # Add a small buffer
    
    try:
        start_time = time.time()
        while (time.time() - start_time < max_duration) and not terminate_program:
            time.sleep(0.1)  # Small sleep to prevent CPU hogging
            
            # Simple progress indicator
            elapsed = time.time() - start_time
            progress = min(elapsed / max_duration, 1.0) * 100
            print(f"\rProgress: {progress:.1f}%", end="")
            
            # Check if both streams have finished their data
            if woojer_position >= len(tactile_data) and looming_position >= len(looming_data):
                break
                
        print("\nPlayback complete")
    except KeyboardInterrupt:
        print("\nPlayback interrupted by user")
    finally:
        # Stop and close streams
        woojer_stream.stop()
        looming_stream.stop()
        woojer_stream.close()
        looming_stream.close()

def main():
    global terminate_program
    
    # List all available devices
    devices = list_audio_devices()
    
    # Try to automatically find the devices
    woojer_id = find_woojer_device(devices)
    looming_id = find_looming_device(devices)
    
    # Allow manual selection if automatic detection fails
    if woojer_id is None:
        print("Woojer device not automatically detected.")
        woojer_id = int(input("Enter ID for Woojer device (for tactile stimuli): "))
    else:
        print(f"Detected Woojer device at ID {woojer_id}: {devices[woojer_id]['name']}")
        print("This device will be used for tactile stimuli.")
    
    if looming_id is None:
        print("Output 1/2 device not automatically detected.")
        print("Available output devices:")
        for i, device in enumerate(devices):
            if device['max_output_channels'] > 0:
                print(f"  {i}: {device['name']}")
        looming_id = int(input("Enter ID for output device (for looming stimuli): "))
    else:
        print(f"Detected output device at ID {looming_id}: {devices[looming_id]['name']}")
        print("This device will be used for looming stimuli.")
    
    # Set the specific file paths
    tactile_file = "C:\\Users\\cogpsy-vrlab\\Documents\\GitHub\\BreathingSpace\\Level1_AudioGeneration\\ExperimentAudio\\participant_0_design_looming.wav"
    looming_file = "C:\\Users\\cogpsy-vrlab\\Documents\\GitHub\\BreathingSpace\\Level1_AudioGeneration\\ExperimentAudio\\participant_0_design_looming.wav"
    
    # Verify files exist
    if not os.path.exists(tactile_file):
        print(f"Warning: Tactile file not found: {tactile_file}")
        tactile_file = input("Please enter the correct path to the tactile file: ")
    
    if not os.path.exists(looming_file):
        print(f"Warning: Looming file not found: {looming_file}")
        looming_file = input("Please enter the correct path to the looming file: ")
    
    # Ask for latency compensation with explanation
    print("\nLatency Compensation:")
    print("This setting helps synchronize the timing between the Woojer device and output device.")
    print("If the tactile stimulus plays with a delay compared to the looming stimulus,")
    print("enter that delay in milliseconds (e.g., 100 for a 100ms delay).")
    print("If you're unsure, start with 0 and adjust as needed after testing.")
    latency_str = input("Enter Woojer latency compensation in milliseconds (default: 0): ")
    latency_compensation = float(latency_str)/1000 if latency_str else 0.0
    
    # Play stimuli
    try:
        play_synchronized_stimuli(woojer_id, looming_id, tactile_file, looming_file, latency_compensation)
        
        # Ask if user wants to play again
        while not terminate_program:
            again = input("Play again? (y/n): ")
            if again.lower() != 'y':
                break
            play_synchronized_stimuli(woojer_id, looming_id, tactile_file, looming_file, latency_compensation)
    except Exception as e:
        print(f"Error during playback: {e}")

if __name__ == "__main__":
    # Initialize global variables
    tactile_data = None
    looming_data = None
    woojer_position = 0
    looming_position = 0
    
    main()
```

```python
import sounddevice as sd
import soundfile as sf
import numpy as np
import time
import os
import matplotlib.pyplot as plt
from scipy import signal

# Parameters for test signal
SAMPLE_RATE = 44100
DURATION = 3  # seconds
PULSE_DURATION = 0.01  # 10ms pulse
PULSE_DELAY = 0.5  # 500ms delay before pulse
RECORDING_EXTRA = 1.0  # Extra recording time to capture delayed signals

def list_audio_devices():
    """List all available audio devices with their IDs"""
    print("Available audio devices:")
    print("-" * 70)
    devices = sd.query_devices()
    for i, device in enumerate(devices):
        input_channels = device.get('max_input_channels', 0)
        output_channels = device.get('max_output_channels', 0)
        name = device.get('name', 'Unknown')
        print(f"ID: {i}, Name: {name}, Inputs: {input_channels}, Outputs: {output_channels}")
    print("-" * 70)
    return devices

def find_device_by_name(devices, name_fragment, exact_match=False, input_device=False):
    """Find a device ID by a partial or exact name match"""
    for i, device in enumerate(devices):
        if exact_match:
            if device['name'] == name_fragment:
                # Check if it's an input or output device as required
                if input_device and device['max_input_channels'] > 0:
                    return i
                elif not input_device and device['max_output_channels'] > 0:
                    return i
        else:
            if name_fragment.lower() in device['name'].lower():
                # Check if it's an input or output device as required
                if input_device and device['max_input_channels'] > 0:
                    return i
                elif not input_device and device['max_output_channels'] > 0:
                    return i
    return None

def find_woojer_device(devices):
    """Find Woojer device for tactile stimuli"""
    for i, device in enumerate(devices):
        if "woojer" in device['name'].lower() and device['max_output_channels'] > 0:
            return i
    return None

def generate_test_signal(sample_rate=SAMPLE_RATE, duration=DURATION, 
                         pulse_duration=PULSE_DURATION, pulse_delay=PULSE_DELAY):
    """Generate a test signal with a sharp pulse for timing measurement"""
    # Create a silent signal
    samples = int(duration * sample_rate)
    signal = np.zeros(samples)
    
    # Add a pulse at the specified delay
    pulse_start = int(pulse_delay * sample_rate)
    pulse_samples = int(pulse_duration * sample_rate)
    pulse_end = pulse_start + pulse_samples
    
    # Create a pulse (using a sine wave at 1000Hz for better recognition)
    t = np.linspace(0, pulse_duration, pulse_samples)
    pulse = np.sin(2 * np.pi * 1000 * t)
    
    # Apply window to the pulse to avoid clicks
    window = np.hamming(len(pulse))
    pulse = pulse * window
    
    # Insert pulse into the signal
    signal[pulse_start:pulse_end] = pulse
    
    return signal, pulse_start / sample_rate

def record_audio(device_id, duration, sample_rate=SAMPLE_RATE, channels=1):
    """Record audio from the specified device"""
    print(f"Recording for {duration:.2f} seconds...")
    recording = sd.rec(
        int(duration * sample_rate),
        samplerate=sample_rate,
        channels=channels,
        device=device_id,
        blocking=True
    )
    return recording

def save_audio(audio_data, filename, sample_rate=SAMPLE_RATE):
    """Save audio data to a WAV file"""
    sf.write(filename, audio_data, sample_rate)
    print(f"Saved audio to {filename}")

def play_audio(device_id, audio_data, sample_rate=SAMPLE_RATE, blocking=True):
    """Play audio on the specified device"""
    # Ensure the data is in the right shape for output
    if len(audio_data.shape) == 1:  # Mono
        data = audio_data.reshape(-1, 1)
    else:
        data = audio_data
    
    # Play the audio
    print(f"Playing audio on device {device_id}...")
    sd.play(data, sample_rate, device=device_id, blocking=blocking)
    if blocking:
        sd.wait()

def detect_pulse_time(audio_data, sample_rate=SAMPLE_RATE, threshold=0.1, window_size=0.01):
    """Detect the time of the first pulse in the audio data"""
    # Apply a high-pass filter to remove low-frequency noise
    sos = signal.butter(10, 500, 'hp', fs=sample_rate, output='sos')
    filtered_data = signal.sosfilt(sos, audio_data)
    
    # Calculate the envelope (absolute value and smoothing)
    window_samples = int(window_size * sample_rate)
    if window_samples % 2 == 0:
        window_samples += 1  # Ensure odd window size for centered smoothing
        
    envelope = np.abs(filtered_data)
    smoothed = signal.savgol_filter(envelope, window_samples, 2)
    
    # Find where the envelope exceeds the threshold
    threshold_value = threshold * np.max(smoothed)
    crossings = np.where(smoothed > threshold_value)[0]
    
    if len(crossings) > 0:
        first_crossing = crossings[0]
        time_seconds = first_crossing / sample_rate
        return time_seconds
    else:
        return None

def plot_signal_analysis(audio_data, pulse_time, expected_time, sample_rate=SAMPLE_RATE, 
                         filename="bluetooth_latency_analysis.png"):
    """Plot audio signal with pulse detection for analysis"""
    fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(12, 10), sharex=True)
    
    # Time axis for plotting
    time_axis = np.arange(len(audio_data)) / sample_rate
    
    # Plot original recording
    ax1.plot(time_axis, audio_data)
    ax1.set_title('Original Recording')
    ax1.set_ylabel('Amplitude')
    ax1.grid(True)
    
    # Apply high-pass filter for clearer visualization
    sos = signal.butter(10, 500, 'hp', fs=sample_rate, output='sos')
    filtered_data = signal.sosfilt(sos, audio_data)
    
    # Plot filtered signal
    ax2.plot(time_axis, filtered_data)
    ax2.set_title('Filtered Signal (High-pass 500Hz)')
    ax2.set_ylabel('Amplitude')
    ax2.grid(True)
    
    # Calculate envelope
    envelope = np.abs(filtered_data)
    window_samples = int(0.01 * sample_rate)
    if window_samples % 2 == 0:
        window_samples += 1
    smoothed = signal.savgol_filter(envelope, window_samples, 2)
    
    # Plot envelope
    ax3.plot(time_axis, smoothed)
    ax3.set_title('Signal Envelope')
    ax3.set_xlabel('Time (s)')
    ax3.set_ylabel('Amplitude')
    ax3.grid(True)
    
    # Mark the expected and detected pulse times on all plots
    for ax in (ax1, ax2, ax3):
        if expected_time:
            ax.axvline(x=expected_time, color='g', linestyle='--', 
                      label=f'Expected pulse at {expected_time:.4f}s')
        if pulse_time:
            ax.axvline(x=pulse_time, color='r', linestyle='-', 
                      label=f'Detected pulse at {pulse_time:.4f}s')
            
    # Add legend to the top plot only
    ax1.legend(loc='upper right')
    
    # Add latency information
    if pulse_time and expected_time:
        latency = pulse_time - expected_time
        fig.suptitle(f'Bluetooth Latency Analysis: {latency * 1000:.2f} ms', fontsize=16)
    else:
        fig.suptitle('Bluetooth Latency Analysis', fontsize=16)
    
    plt.tight_layout()
    plt.savefig(filename)
    print(f"Analysis plot saved to {filename}")
    
    # Try to display the plot if running in an environment that supports it
    try:
        plt.show()
    except:
        pass

def measure_bluetooth_latency():
    """Measure latency of the Bluetooth (Woojer) device"""
    # List all available devices
    devices = list_audio_devices()
    
    # Find Woojer device
    woojer_id = find_woojer_device(devices)
    
    # Find input device (microphone)
    mic_id = find_device_by_name(devices, "microphone", input_device=True)
    
    # Allow manual selection if automatic detection fails
    if woojer_id is None:
        print("Woojer device not automatically detected.")
        print("Available output devices:")
        for i, device in enumerate(devices):
            if device['max_output_channels'] > 0:
                print(f"  {i}: {device['name']}")
        woojer_id = int(input("Enter ID for Woojer device: "))
    else:
        print(f"Detected Woojer device at ID {woojer_id}: {devices[woojer_id]['name']}")
        
    if mic_id is None:
        print("Microphone not automatically detected.")
        print("Available input devices:")
        for i, device in enumerate(devices):
            if device['max_input_channels'] > 0:
                print(f"  {i}: {device['name']}")
        mic_id = int(input("Enter ID for microphone: "))
    else:
        print(f"Detected microphone at ID {mic_id}: {devices[mic_id]['name']}")
    
    # Generate test signal with pulse
    test_signal, expected_pulse_time = generate_test_signal()
    print(f"Generated test signal with pulse at {expected_pulse_time:.4f} seconds")
    
    # Ensure recording directory exists
    os.makedirs("latency_tests", exist_ok=True)
    
    # Testing setup instructions
    print("\n=== BLUETOOTH LATENCY TEST ===")
    print("1. Place the microphone near the Woojer device")
    print("2. Ensure there is minimal background noise")
    print("3. The script will play a short pulse and record it through the microphone")
    input("Press Enter to begin the test...")
    
    # Start recording before playback
    print("Starting recording...")
    recording_thread = threading.Thread(
        target=lambda: record_audio(mic_id, DURATION + RECORDING_EXTRA),
        daemon=True
    )
    recording_thread.start()
    
    # Wait a moment to ensure recording has started
    time.sleep(0.5)
    
    # Play the test signal on Woojer
    play_time = time.time()
    play_audio(woojer_id, test_signal, blocking=True)
    
    # Wait for recording to complete
    wait_time = (DURATION + RECORDING_EXTRA) - (time.time() - play_time)
    if wait_time > 0:
        print(f"Waiting {wait_time:.2f} seconds for recording to complete...")
        time.sleep(wait_time)
    
    # Save the recording
    woojer_recording = recording_thread.join()
    save_audio(woojer_recording, "latency_tests/bluetooth_recording.wav")
    
    # Detect the pulse time in the recording
    woojer_pulse_time = detect_pulse_time(woojer_recording[:, 0])
    if woojer_pulse_time:
        woojer_latency = woojer_pulse_time - expected_pulse_time
        print(f"Bluetooth pulse detected at {woojer_pulse_time:.4f} seconds")
        print(f"Bluetooth latency: {woojer_latency * 1000:.2f} ms")
    else:
        print("No pulse detected in Bluetooth recording")
        woojer_latency = None
    
    # Plot the analysis
    plot_signal_analysis(
        woojer_recording[:, 0],
        woojer_pulse_time,
        expected_pulse_time,
        filename="latency_tests/bluetooth_latency_analysis.png"
    )
    
    # Save the results to a text file
    with open("latency_tests/bluetooth_results.txt", "w") as f:
        f.write("=== BLUETOOTH LATENCY TEST RESULTS ===\n")
        f.write(f"Test time: {time.strftime('%Y-%m-%d %H:%M:%S')}\n\n")
        
        f.write(f"Bluetooth device: {devices[woojer_id]['name']}\n")
        f.write(f"Microphone: {devices[mic_id]['name']}\n\n")
        
        f.write(f"Expected pulse time: {expected_pulse_time:.4f} seconds\n\n")
        
        if woojer_latency is not None:
            f.write(f"Bluetooth pulse detected at: {woojer_pulse_time:.4f} seconds\n")
            f.write(f"Bluetooth latency: {woojer_latency * 1000:.2f} ms\n\n")
            f.write(f"Recommended compensation value: {woojer_latency * 1000:.0f} ms\n")
        else:
            f.write("No pulse detected in Bluetooth recording\n\n")
    
    print(f"\nDetailed results saved to latency_tests/bluetooth_results.txt")
    return woojer_latency

if __name__ == "__main__":
    try:
        measure_bluetooth_latency()
    except Exception as e:
        print(f"Error during latency measurement: {e}")
```

```python
import sounddevice as sd
import soundfile as sf
import numpy as np
import time
import os
import matplotlib.pyplot as plt
from scipy import signal
import queue

# Parameters for test signal
SAMPLE_RATE = 44100
DURATION = 3  # seconds
PULSE_DURATION = 0.01  # 10ms pulse
PULSE_DELAY = 0.5  # 500ms delay before pulse
RECORDING_EXTRA = 1.0  # Extra recording time to capture delayed signals

def list_audio_devices():
    """List all available audio devices with their IDs"""
    print("Available audio devices:")
    print("-" * 70)
    devices = sd.query_devices()
    for i, device in enumerate(devices):
        input_channels = device.get('max_input_channels', 0)
        output_channels = device.get('max_output_channels', 0)
        name = device.get('name', 'Unknown')
        print(f"ID: {i}, Name: {name}, Inputs: {input_channels}, Outputs: {output_channels}")
    print("-" * 70)
    return devices

def find_device_by_name(devices, name_fragment, exact_match=False, input_device=False):
    """Find a device ID by a partial or exact name match"""
    for i, device in enumerate(devices):
        if exact_match:
            if device['name'] == name_fragment:
                # Check if it's an input or output device as required
                if input_device and device['max_input_channels'] > 0:
                    return i
                elif not input_device and device['max_output_channels'] > 0:
                    return i
        else:
            if name_fragment.lower() in device['name'].lower():
                # Check if it's an input or output device as required
                if input_device and device['max_input_channels'] > 0:
                    return i
                elif not input_device and device['max_output_channels'] > 0:
                    return i
    return None

def find_woojer_device(devices):
    """Find Woojer device for tactile stimuli"""
    for i, device in enumerate(devices):
        if "woojer" in device['name'].lower() and device['max_output_channels'] > 0:
            return i
    return None

def generate_test_signal(sample_rate=SAMPLE_RATE, duration=DURATION, 
                         pulse_duration=PULSE_DURATION, pulse_delay=PULSE_DELAY):
    """Generate a test signal with a sharp pulse for timing measurement"""
    # Create a silent signal
    samples = int(duration * sample_rate)
    signal = np.zeros(samples)
    
    # Add a pulse at the specified delay
    pulse_start = int(pulse_delay * sample_rate)
    pulse_samples = int(pulse_duration * sample_rate)
    pulse_end = pulse_start + pulse_samples
    
    # Create a pulse (using a sine wave at 1000Hz for better recognition)
    t = np.linspace(0, pulse_duration, pulse_samples)
    pulse = np.sin(2 * np.pi * 1000 * t)
    
    # Apply window to the pulse to avoid clicks
    window = np.hamming(len(pulse))
    pulse = pulse * window
    
    # Insert pulse into the signal
    signal[pulse_start:pulse_end] = pulse
    
    return signal, pulse_start / sample_rate

def record_audio(device_id, duration, sample_rate=SAMPLE_RATE, channels=1):
    """Record audio from the specified device using blocking mode"""
    print(f"Recording for {duration:.2f} seconds...")
    recording = sd.rec(
        int(duration * sample_rate),
        samplerate=sample_rate,
        channels=channels,
        device=device_id,
        blocking=True
    )
    return recording

def save_audio(audio_data, filename, sample_rate=SAMPLE_RATE):
    """Save audio data to a WAV file"""
    sf.write(filename, audio_data, sample_rate)
    print(f"Saved audio to {filename}")

def play_audio(device_id, audio_data, sample_rate=SAMPLE_RATE, blocking=True):
    """Play audio on the specified device"""
    # Ensure the data is in the right shape for output
    if len(audio_data.shape) == 1:  # Mono
        data = audio_data.reshape(-1, 1)
    else:
        data = audio_data
    
    # Play the audio
    print(f"Playing audio on device {device_id}...")
    sd.play(data, sample_rate, device=device_id, blocking=blocking)
    if blocking:
        sd.wait()

def detect_pulse_time(audio_data, sample_rate=SAMPLE_RATE, threshold=0.1, window_size=0.01):
    """Detect the time of the first pulse in the audio data"""
    # Apply a high-pass filter to remove low-frequency noise
    sos = signal.butter(10, 500, 'hp', fs=sample_rate, output='sos')
    filtered_data = signal.sosfilt(sos, audio_data)
    
    # Calculate the envelope (absolute value and smoothing)
    window_samples = int(window_size * sample_rate)
    if window_samples % 2 == 0:
        window_samples += 1  # Ensure odd window size for centered smoothing
        
    envelope = np.abs(filtered_data)
    smoothed = signal.savgol_filter(envelope, window_samples, 2)
    
    # Find where the envelope exceeds the threshold
    threshold_value = threshold * np.max(smoothed)
    crossings = np.where(smoothed > threshold_value)[0]
    
    if len(crossings) > 0:
        first_crossing = crossings[0]
        time_seconds = first_crossing / sample_rate
        return time_seconds
    else:
        return None

def plot_signal_analysis(audio_data, pulse_time, expected_time, sample_rate=SAMPLE_RATE, 
                         filename="bluetooth_latency_analysis.png"):
    """Plot audio signal with pulse detection for analysis"""
    fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(12, 10), sharex=True)
    
    # Time axis for plotting
    time_axis = np.arange(len(audio_data)) / sample_rate
    
    # Plot original recording
    ax1.plot(time_axis, audio_data)
    ax1.set_title('Original Recording')
    ax1.set_ylabel('Amplitude')
    ax1.grid(True)
    
    # Apply high-pass filter for clearer visualization
    sos = signal.butter(10, 500, 'hp', fs=sample_rate, output='sos')
    filtered_data = signal.sosfilt(sos, audio_data)
    
    # Plot filtered signal
    ax2.plot(time_axis, filtered_data)
    ax2.set_title('Filtered Signal (High-pass 500Hz)')
    ax2.set_ylabel('Amplitude')
    ax2.grid(True)
    
    # Calculate envelope
    envelope = np.abs(filtered_data)
    window_samples = int(0.01 * sample_rate)
    if window_samples % 2 == 0:
        window_samples += 1
    smoothed = signal.savgol_filter(envelope, window_samples, 2)
    
    # Plot envelope
    ax3.plot(time_axis, smoothed)
    ax3.set_title('Signal Envelope')
    ax3.set_xlabel('Time (s)')
    ax3.set_ylabel('Amplitude')
    ax3.grid(True)
    
    # Mark the expected and detected pulse times on all plots
    for ax in (ax1, ax2, ax3):
        if expected_time:
            ax.axvline(x=expected_time, color='g', linestyle='--', 
                      label=f'Expected pulse at {expected_time:.4f}s')
        if pulse_time:
            ax.axvline(x=pulse_time, color='r', linestyle='-', 
                      label=f'Detected pulse at {pulse_time:.4f}s')
            
    # Add legend to the top plot only
    ax1.legend(loc='upper right')
    
    # Add latency information
    if pulse_time and expected_time:
        latency = pulse_time - expected_time
        fig.suptitle(f'Bluetooth Latency Analysis: {latency * 1000:.2f} ms', fontsize=16)
    else:
        fig.suptitle('Bluetooth Latency Analysis', fontsize=16)
    
    plt.tight_layout()
    plt.savefig(filename)
    print(f"Analysis plot saved to {filename}")
    
    # Try to display the plot if running in an environment that supports it
    try:
        plt.show()
    except:
        pass

def measure_bluetooth_latency():
    """Measure latency of the Bluetooth (Woojer) device"""
    # List all available devices
    devices = list_audio_devices()
    
    # Find Woojer device
    woojer_id = find_woojer_device(devices)
    
    # Find input device (microphone)
    mic_id = find_device_by_name(devices, "microphone", input_device=True)
    if mic_id is None:
        # Try to find any input from Woojer
        for i, device in enumerate(devices):
            if "woojer" in device['name'].lower() and device['max_input_channels'] > 0:
                mic_id = i
                break
    
    # Allow manual selection if automatic detection fails
    if woojer_id is None:
        print("Woojer device not automatically detected.")
        print("Available output devices:")
        for i, device in enumerate(devices):
            if device['max_output_channels'] > 0:
                print(f"  {i}: {device['name']}")
        woojer_id = int(input("Enter ID for Woojer device: "))
    else:
        print(f"Detected Woojer device at ID {woojer_id}: {devices[woojer_id]['name']}")
        
    if mic_id is None:
        print("Input device not automatically detected.")
        print("Available input devices:")
        for i, device in enumerate(devices):
            if device['max_input_channels'] > 0:
                print(f"  {i}: {device['name']}")
        mic_id = int(input("Enter ID for input device: "))
    else:
        print(f"Detected input device at ID {mic_id}: {devices[mic_id]['name']}")
    
    # Generate test signal with pulse
    test_signal, expected_pulse_time = generate_test_signal()
    print(f"Generated test signal with pulse at {expected_pulse_time:.4f} seconds")
    
    # Ensure recording directory exists
    os.makedirs("latency_tests", exist_ok=True)
    
    # Testing setup instructions
    print("\n=== BLUETOOTH LATENCY TEST ===")
    print("1. Connect the Woojer output to your PC's input")
    print("2. Ensure there is minimal background noise")
    print("3. The script will play a short pulse through Woojer and record it through the input")
    input("Press Enter to begin the test...")
    
    # First record, then play (separate operations to avoid thread issues)
    print("Starting recording...")
    
    # Create a separate thread for recording
    q = queue.Queue()
    
    def recording_callback(indata, frames, time, status):
        if status:
            print(f"Recording status: {status}")
        q.put(indata.copy())
    
    # Setup recording stream
    stream = sd.InputStream(
        device=mic_id,
        channels=1,
        samplerate=SAMPLE_RATE,
        callback=recording_callback
    )
    
    # Start recording
    with stream:
        # Play test signal
        play_audio(woojer_id, test_signal, blocking=False)
        
        # Record for specified duration
        print(f"Recording for {DURATION + RECORDING_EXTRA:.2f} seconds...")
        recorded_chunks = []
        start_time = time.time()
        
        while time.time() - start_time < DURATION + RECORDING_EXTRA:
            try:
                chunk = q.get(timeout=0.1)
                recorded_chunks.append(chunk)
            except queue.Empty:
                pass
    
    # Process recorded data
    if recorded_chunks:
        recording = np.vstack(recorded_chunks)
        
        # Save the recording
        save_audio(recording, "latency_tests/bluetooth_recording.wav")
        
        # Detect the pulse time in the recording
        bluetooth_pulse_time = detect_pulse_time(recording[:, 0])
        if bluetooth_pulse_time:
            bluetooth_latency = bluetooth_pulse_time - expected_pulse_time
            print(f"Bluetooth pulse detected at {bluetooth_pulse_time:.4f} seconds")
            print(f"Bluetooth latency: {bluetooth_latency * 1000:.2f} ms")
        else:
            print("No pulse detected in Bluetooth recording")
            bluetooth_latency = None
        
        # Plot the analysis
        plot_signal_analysis(
            recording[:, 0],
            bluetooth_pulse_time,
            expected_pulse_time,
            filename="latency_tests/bluetooth_latency_analysis.png"
        )
        
        # Save the results to a text file
        with open("latency_tests/bluetooth_results.txt", "w") as f:
            f.write("=== BLUETOOTH LATENCY TEST RESULTS ===\n")
            f.write(f"Test time: {time.strftime('%Y-%m-%d %H:%M:%S')}\n\n")
            
            f.write(f"Bluetooth device: {devices[woojer_id]['name']}\n")
            f.write(f"Input device: {devices[mic_id]['name']}\n\n")
            
            f.write(f"Expected pulse time: {expected_pulse_time:.4f} seconds\n\n")
            
            if bluetooth_latency is not None:
                f.write(f"Bluetooth pulse detected at: {bluetooth_pulse_time:.4f} seconds\n")
                f.write(f"Bluetooth latency: {bluetooth_latency * 1000:.2f} ms\n\n")
                f.write(f"Recommended compensation value: {bluetooth_latency * 1000:.0f} ms\n")
            else:
                f.write("No pulse detected in Bluetooth recording\n\n")
        
        print(f"\nDetailed results saved to latency_tests/bluetooth_results.txt")
        return bluetooth_latency
    else:
        print("Error: No audio was recorded")
        return None

if __name__ == "__main__":
    try:
        measure_bluetooth_latency()
    except Exception as e:
        print(f"Error during latency measurement: {e}")
        import traceback
        traceback.print_exc()
```

```python
import sounddevice as sd
import soundfile as sf
import numpy as np
import time
import os
import queue
import threading
from datetime import datetime

# Parameters
SAMPLE_RATE = 44100
MAX_DURATION = 60  # Play first 60 seconds
RECORDING_BUFFER = 2  # Extra seconds to record after playback ends

def list_audio_devices():
    """List all available audio devices with their IDs"""
    print("Available audio devices:")
    print("-" * 70)
    devices = sd.query_devices()
    for i, device in enumerate(devices):
        input_channels = device.get('max_input_channels', 0)
        output_channels = device.get('max_output_channels', 0)
        name = device.get('name', 'Unknown')
        print(f"ID: {i}, Name: {name}, Inputs: {input_channels}, Outputs: {output_channels}")
    print("-" * 70)
    return devices

def find_woojer_device(devices):
    """Find Woojer device for tactile stimuli"""
    for i, device in enumerate(devices):
        if "woojer" in device['name'].lower() and device['max_output_channels'] > 0:
            return i
    return None

def find_all_input_devices(devices):
    """Find all microphone/input devices"""
    input_devices = []
    for i, device in enumerate(devices):
        if device.get('max_input_channels', 0) > 0:
            input_devices.append({
                'id': i,
                'name': device.get('name', 'Unknown'),
                'channels': device.get('max_input_channels', 0)
            })
    return input_devices

def record_from_device(device_id, device_name, duration, sample_rate=SAMPLE_RATE, channels=1):
    """Record from a specific input device and save to a file"""
    downloads_folder = os.path.join(os.path.expanduser("~"), "Downloads")
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    filename = os.path.join(downloads_folder, f"device_{device_id}_{device_name.replace(' ', '_')}_{timestamp}.wav")
    
    # Create a queue for the recorded audio data
    q = queue.Queue()
    
    # Flag to signal when to stop
    recording_stopped = threading.Event()
    
    # Callback function for the input stream
    def callback(indata, frames, time, status):
        if status:
            print(f"Recording status for {device_name}: {status}")
        q.put(indata.copy())
    
    # Process to save the recording to a file
    def save_recording():
        try:
            # Create an input stream for this device
            with sd.InputStream(
                device=device_id,
                channels=min(channels, 2),  # Limit to 2 channels for WAV file compatibility
                samplerate=sample_rate,
                callback=callback
            ):
                print(f"Started recording from {device_name} (ID: {device_id})")
                
                # Collect data from the queue until recording_stopped is set
                recorded_data = []
                while not recording_stopped.is_set() or not q.empty():
                    try:
                        data = q.get(timeout=0.1)
                        recorded_data.append(data)
                    except queue.Empty:
                        if recording_stopped.is_set() and q.empty():
                            break
                
                if recorded_data:
                    # Combine all chunks
                    recording = np.vstack(recorded_data)
                    
                    # Save to file
                    sf.write(filename, recording, sample_rate)
                    print(f"Recording from {device_name} saved to: {filename}")
                else:
                    print(f"No data recorded from {device_name}")
        except Exception as e:
            print(f"Error recording from {device_name}: {e}")
    
    # Start the recording process in a separate thread
    thread = threading.Thread(target=save_recording)
    thread.daemon = True
    thread.start()
    
    # Return the thread and event for controlling the recording
    return thread, recording_stopped, filename

def play_and_record_all_mics():
    """Play looming stimulus through Woojer and record from all microphones simultaneously"""
    # List all available devices
    devices = list_audio_devices()
    
    # Find Woojer device
    woojer_id = find_woojer_device(devices)
    
    # Find all input devices
    input_devices = find_all_input_devices(devices)
    
    # Allow manual selection if automatic detection fails
    if woojer_id is None:
        print("Woojer device not automatically detected.")
        print("Available output devices:")
        for i, device in enumerate(devices):
            if device.get('max_output_channels', 0) > 0:
                print(f"  {i}: {device['name']}")
        woojer_id = int(input("Enter ID for Woojer device: "))
    else:
        print(f"Detected Woojer device at ID {woojer_id}: {devices[woojer_id]['name']}")
    
    if not input_devices:
        print("No input devices found.")
        return
    
    print(f"Found {len(input_devices)} input devices:")
    for idx, device in enumerate(input_devices):
        print(f"  {idx+1}. ID: {device['id']}, Name: {device['name']}, Channels: {device['channels']}")
    
    # Ask which input devices to use (default is all)
    use_all = input("Record from all input devices? (y/n, default y): ")
    
    if use_all.lower() == 'n':
        selected_indices = input("Enter the numbers of devices to use (comma separated): ")
        selected_indices = [int(idx.strip()) - 1 for idx in selected_indices.split(',')]
        selected_devices = [input_devices[idx] for idx in selected_indices if 0 <= idx < len(input_devices)]
    else:
        selected_devices = input_devices
    
    # Load the looming stimulus
    looming_file = "C:\\Users\\cogpsy-vrlab\\Documents\\GitHub\\BreathingSpace\\Level1_AudioGeneration\\ExperimentAudio\\participant_0_design_looming.wav"
    
    # Check if file exists
    if not os.path.exists(looming_file):
        print(f"Looming stimulus file not found: {looming_file}")
        looming_file = input("Please enter the full path to the looming stimulus file: ")
    
    # Load audio
    print(f"Loading looming stimulus: {looming_file}")
    looming_data, sample_rate = sf.read(looming_file)
    
    # Trim to MAX_DURATION if needed
    if len(looming_data) > MAX_DURATION * sample_rate:
        print(f"Trimming stimulus to first {MAX_DURATION} seconds")
        looming_data = looming_data[:int(MAX_DURATION * sample_rate)]
    
    # Get actual duration
    actual_duration = len(looming_data) / sample_rate
    print(f"Stimulus duration: {actual_duration:.2f} seconds")
    
    # Prepare output data for the Woojer
    if len(looming_data.shape) == 1:  # If mono
        woojer_output = looming_data.reshape(-1, 1)
    else:  # If stereo, use left channel for simplicity
        woojer_output = looming_data[:, 0].reshape(-1, 1)
    
    # Normalize to maximum volume for better detection
    woojer_output = woojer_output / np.max(np.abs(woojer_output)) * 0.9
    
    # Start recording from all selected input devices
    recording_threads = []
    recording_stopped_events = []
    output_files = []
    
    print("\n=== STARTING MULTI-MICROPHONE TEST ===")
    print(f"Starting recordings from {len(selected_devices)} microphones")
    print(f"All recordings will be saved to your Downloads folder\n")
    
    for device in selected_devices:
        thread, stopped_event, output_file = record_from_device(
            device['id'], 
            device['name'].replace('/', '_').replace('\\', '_'),  # Sanitize name for filename
            actual_duration + RECORDING_BUFFER,
            channels=device['channels']
        )
        recording_threads.append(thread)
        recording_stopped_events.append(stopped_event)
        output_files.append(output_file)
    
    # Wait a moment for all recordings to start
    time.sleep(1)
    
    # Play through Woojer
    print(f"Playing looming stimulus through Woojer for {actual_duration:.2f} seconds...")
    start_time = time.time()
    sd.play(woojer_output, sample_rate, device=woojer_id, blocking=True)
    end_time = time.time()
    
    # Add a small buffer to ensure we capture everything
    buffer_time = min(RECORDING_BUFFER, 2.0)  # Max 2 seconds buffer
    print(f"Playback complete. Continuing to record for {buffer_time:.1f} more seconds...")
    time.sleep(buffer_time)
    
    # Signal all recordings to stop
    for stopped_event in recording_stopped_events:
        stopped_event.set()
    
    # Wait for all recording threads to finish
    for thread in recording_threads:
        thread.join()
    
    print("\n=== TEST COMPLETE ===")
    print("All recordings saved to your Downloads folder:")
    for output_file in output_files:
        print(f"  - {os.path.basename(output_file)}")
    
    return output_files

if __name__ == "__main__":
    try:
        play_and_record_all_mics()
    except Exception as e:
        print(f"Error during test: {e}")
        import traceback
        traceback.print_exc()
```

```python
import sounddevice as sd
import soundfile as sf
import numpy as np
import time
import os

# Parameters
SAMPLE_RATE = 44100
MAX_VOLUME = 0.9  # Set to 90% of maximum to avoid distortion

def list_audio_devices():
    """List all available audio devices with their IDs"""
    print("Available audio devices:")
    print("-" * 70)
    devices = sd.query_devices()
    for i, device in enumerate(devices):
        input_channels = device.get('max_input_channels', 0)
        output_channels = device.get('max_output_channels', 0)
        name = device.get('name', 'Unknown')
        print(f"ID: {i}, Name: {name}, Inputs: {input_channels}, Outputs: {output_channels}")
    print("-" * 70)
    return devices

def find_woojer_devices(devices):
    """Find all possible Woojer output devices"""
    woojer_devices = []
    for i, device in enumerate(devices):
        if "woojer" in device['name'].lower() and device['max_output_channels'] > 0:
            woojer_devices.append((i, device['name']))
    return woojer_devices

def play_looming_through_woojer():
    """Play looming stimulus through Woojer at maximum volume"""
    # List all available devices
    devices = list_audio_devices()
    
    # Find all Woojer devices
    woojer_devices = find_woojer_devices(devices)
    
    if not woojer_devices:
        print("No Woojer devices detected automatically.")
        print("Available output devices:")
        for i, device in enumerate(devices):
            if device.get('max_output_channels', 0) > 0:
                print(f"  {i}: {device['name']}")
        woojer_id = int(input("Enter ID for Woojer device: "))
    elif len(woojer_devices) == 1:
        woojer_id = woojer_devices[0][0]
        print(f"Detected Woojer device: {woojer_devices[0][1]} (ID: {woojer_id})")
    else:
        print("Multiple Woojer devices detected:")
        for i, (device_id, device_name) in enumerate(woojer_devices):
            print(f"  {i+1}. {device_name} (ID: {device_id})")
        selection = int(input("Select which Woojer device to use (1, 2, etc.): "))
        woojer_id = woojer_devices[selection-1][0]
    
    # Load the looming stimulus
    looming_file = "C:\\Users\\cogpsy-vrlab\\Documents\\GitHub\\BreathingSpace\\Level1_AudioGeneration\\ExperimentAudio\\participant_0_design_looming.wav"
    
    # Check if file exists
    if not os.path.exists(looming_file):
        print(f"Looming stimulus file not found: {looming_file}")
        looming_file = input("Please enter the full path to the looming stimulus file: ")
    
    # Load audio
    print(f"Loading looming stimulus: {looming_file}")
    looming_data, sample_rate = sf.read(looming_file)
    
    # Get duration
    duration = len(looming_data) / sample_rate
    print(f"Stimulus duration: {duration:.2f} seconds")
    
    # Prepare output data for the Woojer (convert to mono if needed)
    if len(looming_data.shape) == 1:  # If mono
        woojer_output = looming_data.reshape(-1, 1)
    else:  # If stereo, convert to mono by averaging channels
        woojer_output = np.mean(looming_data, axis=1).reshape(-1, 1)
    
    # Normalize to maximum volume
    max_amplitude = np.max(np.abs(woojer_output))
    if max_amplitude > 0:
        woojer_output = woojer_output / max_amplitude * MAX_VOLUME
    
    # Try different devices if first attempt fails
    if len(woojer_devices) > 1:
        print("\nWill try multiple Woojer devices if needed.")
        print("Press Ctrl+C to stop and try the next device.")
    
    # Try each Woojer device
    for attempt, (device_id, device_name) in enumerate(woojer_devices):
        if attempt > 0:
            choice = input(f"\nTry next device? ({device_name}, ID: {device_id}) (y/n): ")
            if choice.lower() != 'y':
                break
        
        try:
            print(f"\nPlaying through {device_name} (ID: {device_id})")
            print(f"Duration: {duration:.2f} seconds")
            print("Press Ctrl+C to stop playback")
            
            # Play through Woojer
            sd.play(woojer_output, sample_rate, device=device_id, blocking=False)
            
            # Wait while playing
            start_time = time.time()
            try:
                while time.time() - start_time < duration:
                    elapsed = time.time() - start_time
                    print(f"\rPlayback progress: {elapsed:.1f}/{duration:.1f} seconds", end="")
                    time.sleep(0.2)
                print("\nPlayback complete!")
            except KeyboardInterrupt:
                sd.stop()
                print("\nPlayback stopped by user")
        
        except Exception as e:
            print(f"Error playing through device {device_id}: {e}")
    
    print("\nTest complete")

if __name__ == "__main__":
    try:
        play_looming_through_woojer()
    except Exception as e:
        print(f"Error: {e}")
        import traceback
        traceback.print_exc()
```
