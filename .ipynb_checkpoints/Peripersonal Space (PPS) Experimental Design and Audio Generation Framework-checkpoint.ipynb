# Peripersonal Space Experimental Framework

## Project Overview

This computational framework is designed to investigate human spatial perception through a sophisticated, controlled experimental paradigm that explores how breathing patterns influence sensory processing and spatial awareness.

## Core Objectives

The script generates a comprehensive experimental setup that:
1. Creates counterbalanced experimental designs for 100 participants
2. Produces precisely synchronized audio stimuli
3. Ensures rigorous scientific methodology through advanced validation techniques

## Key Parameters

### Experimental Design
- **Participants**: 100 unique designs
- **Total Trials**: 202 trials per participant
  * Inhalation Trials: 60
  * Exhalation Trials: 60
  * Baseline Trials: 60
  * Catch Trials: 22

### Stimulus Characteristics
- **Spatial Audio Stimuli**: 
  * Right azimuth (90 degrees)
  * Front azimuth (0 degrees)
  * Left azimuth (-90 degrees)

### Timing Parameters
- **Stimulus Onset Asynchrony (SOA)**: 5 conditions
  * 190 ms
  * 400 ms
  * 700 ms
  * 1000 ms
  * 1500 ms

- **Temporal Jitter**: Random delays
  * 100 ms
  * 200 ms
  * 300 ms
  * 400 ms
  * 500 ms

## Computational Workflow

### Design Generation
- Generates counterbalanced experimental designs
- Ensures balanced distribution of:
  * Stimulus types
  * Trial types
  * SOA conditions
  * Temporal jitter

### Validation Techniques
- Doegen-inspired statistical checks
- Comprehensive validation including:
  * Transition matrix analysis
  * Factor orthogonality
  * Sequence randomness testing
  * Autocorrelation analysis

### Audio Generation
- Creates two synchronized audio files per participant:
  * Looming stimuli embedded in box breathing audio
  * Tactile stimuli track with precise SOA adjustments

```python
# ======================================================================
# COMBINED COUNTERBALANCED DESIGN AND AUDIO GENERATOR FOR PPS EXPERIMENT
# ======================================================================
# 
# This script integrates two key components of a Peripersonal Space (PPS) experiment:
# 1. A counterbalancer that generates experimental designs with balanced conditions
# 2. An audio generator that creates TWO participant-specific audio files with embedded stimuli:
#    - First audio: Box breathing with embedded directional looming stimuli
#    - Second audio: Synchronized tactile stimuli with applied SOA offsets
#
# PROJECT OVERVIEW:
# This experiment investigates how breathing phases (inhalation/exhalation) influence 
# responses to looming stimuli in peripersonal space. The design presents auditory stimuli 
# from different directions (left/front/right) during controlled breathing exercises.
# Each looming stimulus is paired with a tactile stimulus with a specific SOA (Stimulus Onset Asynchrony).
#
# VALIDATION:
# This script includes comprehensive counterbalancing validation using Doegen-inspired
# validation checks to ensure proper experimental design balance across all factors.
"""
Peripersonal Space (PPS) Audio Stimuli Generator

PROJECT OVERVIEW:
This script is part of a cognitive neuroscience experiment investigating peripersonal space (PPS) 
perception during controlled breathing exercises. PPS refers to the space immediately surrounding 
the body that is integrated with tactile perception. The experiment aims to understand how PPS 
representation may change during different phases of breathing.

PURPOSE:
This script generates participant-specific audio files by injecting spatial audio stimuli 
(sounds that appear to come from specific directions) into a guided box breathing meditation 
audio track. Each participant receives a unique audio file where stimuli are presented at 
specific breathing hold phases with controlled randomization (jitter).

EXPERIMENTAL DESIGN:
1. Participants follow a guided box breathing meditation (inhale, hold, exhale, hold)
2. During specific hold phases, 3D spatial audio stimuli are presented from different directions
3. The timing of stimuli is precisely controlled but includes small random variations (jitter)
   to prevent anticipation effects
4. Each participant's design specifies which type of spatial audio stimuli (left/front/right)
   should be presented during which breathing cycles

STIMULI DESCRIPTION:
- Base audio: A guided box breathing meditation instructing participants when to inhale, 
  hold, exhale, and hold
- Spatial stimuli: 3D audio recordings using FABIAN HRIR (Head-Related Impulse Response) that
  create perception of sounds approaching from three directions:
  * Right (90° azimuth)
  * Front (0° azimuth) 
  * Left (-90° azimuth)

DATA FLOW:
1. Load participant-specific design CSV (specifying which trial gets which stimulus)
2. Load base box breathing audio (meditation instructions)
3. For each trial in the design:
   a. Determine exact sample position based on reference timestamps
   b. Add controlled randomization (jitter)
   c. Insert appropriate spatial audio stimulus
4. Save personalized audio file and detailed logs of stimulus timing

EXPECTED USE:
The generated audio files will be used in experiments where participants:
1. Listen to the personalized box breathing audio through headphones
2. Follow the breathing instructions
3. Experience the spatial audio stimuli during specific breathing phases
4. Potentially respond to these stimuli (reaction time measurements)
5. Data will help understand how PPS perception might change during different 
   breathing phases

AUTHOR: George Fejer
"""
import os
import tkinter as tk
from tkinter import simpledialog
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import soundfile as sf
import scipy.signal as signal
from datetime import datetime
from itertools import product, combinations
import warnings
import json

# ======================================================================
# USER-CONFIGURABLE PARAMETERS
# ======================================================================

# SOA conditions in milliseconds (Stimulus Onset Asynchrony)
# Positive values mean tactile comes after looming, negative means tactile comes before
SOA_CONDITIONS_MS = [190, 400, 700, 1000, 1500]  # Replace with your desired values

# File paths - UPDATE THESE TO MATCH YOUR SYSTEM
BASE_DIR = r"C:\Users\cogpsy-vrlab\Documents\PPS_module\BreathingPilot"

# Box breathing audio
BASE_AUDIO_PATH = os.path.join(BASE_DIR, "AudioInstructionsBoxBreathing", "TargetStimuli", 
                              "6. Box Breathing Combined (Segment Injected).wav")

# Timestamp reference for breathing phases
TIMESTAMPS_CSV_PATH = os.path.join(BASE_DIR, "AudioInstructionsBoxBreathing", "TargetStimuli", 
                                  "BoxBreathing_extended_timestamps.csv")

# Output directories
DESIGN_OUTPUT_DIR = os.path.join(BASE_DIR, "PPS_Experiment_Module", "ExperimentLog")
AUDIO_OUTPUT_DIR = os.path.join(BASE_DIR, "PPS_Experiment_Module", "ExperimentAudio")

# Looming stimuli (directional audio)
LOOMING_STIMULI = {
    'right': os.path.join(BASE_DIR, "PPS_Experiment_Module", "right_az90_FABIAN_HRIR_natural.wav"),
    'front': os.path.join(BASE_DIR, "PPS_Experiment_Module", "front_az0_FABIAN_HRIR_natural.wav"),
    'left': os.path.join(BASE_DIR, "PPS_Experiment_Module", "left_az-90_FABIAN_HRIR_natural.wav")
}

# Tactile stimulus (to be synchronized with looming stimuli + SOA)
TACTILE_STIMULUS_PATH = os.path.join(BASE_DIR, "PPS_Experiment_Module", "tactile_stimulus.wav")

# Number of participants to generate designs for
NUM_PARTICIPANTS = 100

# Trial counts for each condition
TRIAL_COUNTS = {
    'inhalation': 60,   # Trials during inhalation phase
    'exhalation': 60,   # Trials during exhalation phase
    'baseline': 60,     # Baseline trials (neutral breathing)
    'catch': 24         # Catch trials to maintain attention
}

# Jitter options in milliseconds (temporal variability)
JITTER_OPTIONS_MS = [100, 200, 300, 400, 500]

# ======================================================================
# COUNTERBALANCED DESIGN GENERATOR
# ======================================================================

class ComprehensiveCounterbalancer:
    """
    Creates counterbalanced experimental designs for multiple participants.
    
    This class generates trial sequences ensuring balanced presentation of 
    experimental factors (stimulus direction, trial type, SOA conditions) 
    across participants. It can use reference timestamps from a breathing exercise
    to align stimuli with specific breathing phases.
    """
    
    def __init__(self, output_dir, soa_conditions, trial_counts, jitter_options):
        """
        Initialize the counterbalancer with experiment parameters.
        
        Args:
            output_dir: Directory to save design files
            soa_conditions: List of SOA values in milliseconds
            trial_counts: Dictionary with number of trials per condition
            jitter_options: List of jitter values in milliseconds
        """
        # Ensure output directory exists
        self.output_dir = output_dir
        os.makedirs(output_dir, exist_ok=True)
        
        # Define experimental factors
        self.factors = {
            'stimulus_type': ['right', 'front', 'left'],  # Stimulus directions
            'trial_type': ['inhalation', 'exhalation', 'baseline', 'catch'],  # Breathing phases
            'soa_condition': list(range(len(soa_conditions))),  # SOA condition indices (0-4)
            'jitter': jitter_options  # Jitter in milliseconds for temporal variability
        }
        
        # Store actual SOA values for reference
        self.soa_values_ms = soa_conditions
        
        # Experiment design parameters - number of trials per condition
        self.trial_counts = trial_counts
        
        # Default timing parameters (used if no reference CSV is provided)
        self.base_time = pd.Timestamp('2025-03-01 09:00:00')  # Starting time for the experiment
        self.trial_duration = 8  # Seconds per trial
        
        # This will be set if a reference CSV is provided
        self.reference_csv_path = None

    def generate_comprehensive_design(self, num_participants=100, reference_csv_path=None):
        """
        Generate a counterbalanced design for multiple participants.
        
        Args:
            num_participants: Number of participant designs to generate
            reference_csv_path: Path to CSV with breathing exercise timestamps
                                (used to align stimuli with breathing phases)
        
        Returns:
            List of DataFrames, one for each participant
        """
        self.reference_csv_path = reference_csv_path
        designs = []
        
        # Set a fixed seed for reproducibility
        np.random.seed(42)
        
        # Generate design for each participant
        for participant_id in range(num_participants):
            design = self._generate_single_participant_design(participant_id)
            designs.append(design)
        
        return designs
    
    def _generate_single_participant_design(self, participant_id):
        """
        Generate a balanced design for a single participant with precise timestamps.
        
        For each trial, computes three timestamps:
          - timestamp_original: baseline timing
          - timestamp_after_jitter: timing with added jitter
          - timestamp_with_soa: timing with jitter and SOA adjustment
        
        Args:
            participant_id: Unique identifier for the participant
        
        Returns:
            DataFrame with complete experimental design including timestamps
        """
        design = []
        
        # If a reference CSV is provided, use it for timing
        if self.reference_csv_path is not None:
            timestamp_df = pd.read_csv(self.reference_csv_path)
            csv_milliseconds = timestamp_df['milliseconds'].astype(int).values
            total_trials = sum(self.trial_counts.values())
            if total_trials > len(csv_milliseconds):
                raise ValueError(f"Not enough timestamps in CSV! Needed {total_trials}, but found {len(csv_milliseconds)} rows.")
            csv_index = 0  # Track which CSV row to use for each trial
        
        # Generate designs for each trial type (inhalation, exhalation, etc.)
        for trial_type, count in self.trial_counts.items():
            # Create a balanced stimulus sequence for each trial type
            stimulus_sequence = []
            for stim_type in self.factors['stimulus_type']:
                # Distribute stimulus types evenly within each trial type
                stimulus_sequence.extend([stim_type] * (count // len(self.factors['stimulus_type'])))
            
            # Shuffle the stimulus sequence for pseudorandomization
            np.random.shuffle(stimulus_sequence)
            
            # Add design details for each trial
            for idx, stimulus_type in enumerate(stimulus_sequence):
                # Compute trial number (starting after an initial offset)
                trial_idx = len(design)
                trial_number = 20 + trial_idx  # Offset of 20 trials
                
                # Choose a random jitter value (in ms)
                current_jitter = np.random.choice(self.factors['jitter'])
                
                # Get SOA condition index and actual value in milliseconds
                soa_idx = idx % len(self.factors['soa_condition'])
                soa_value_ms = self.soa_values_ms[soa_idx]
                
                # Generate timestamps (either from CSV or calculated)
                if self.reference_csv_path is not None:
                    # Use CSV-based timestamps (linked to breathing audio)
                    base_ms = csv_milliseconds[csv_index]
                    csv_index += 1
                    jittered_ms = base_ms + current_jitter
                    soa_ms = jittered_ms + soa_value_ms
                    
                    # Format timestamps as "mm:ss.f"
                    timestamp_original = f"{int(base_ms // 60000):02}:{(base_ms % 60000) / 1000:.1f}"
                    timestamp_after_jitter = f"{int(jittered_ms // 60000):02}:{(jittered_ms % 60000) / 1000:.1f}"
                    timestamp_with_soa = f"{int(soa_ms // 60000):02}:{(soa_ms % 60000) / 1000:.1f}"
                    
                    # Store raw millisecond values for precise audio syncing
                    base_ms_raw = base_ms
                    jittered_ms_raw = jittered_ms
                    soa_ms_raw = soa_ms
                else:
                    # Use base_time and trial_duration method (regular intervals)
                    t_orig = self.base_time + pd.Timedelta(seconds=trial_number * self.trial_duration)
                    t_jitter = self.base_time + pd.Timedelta(seconds=(trial_number * self.trial_duration) + (current_jitter / 1000))
                    t_soa = self.base_time + pd.Timedelta(seconds=(trial_number * self.trial_duration) + (current_jitter / 1000) + (soa_value_ms / 1000))
                    
                    timestamp_original = t_orig.strftime('%M:%S.%f')[:-4]
                    timestamp_after_jitter = t_jitter.strftime('%M:%S.%f')[:-4]
                    timestamp_with_soa = t_soa.strftime('%M:%S.%f')[:-4]
                    
                    # Calculate raw millisecond values for precise audio syncing
                    base_ms_raw = trial_number * self.trial_duration * 1000
                    jittered_ms_raw = base_ms_raw + current_jitter
                    soa_ms_raw = jittered_ms_raw + soa_value_ms
                
                # Add this trial to the design
                design.append({
                    'participant_id': participant_id,
                    'trial_number': trial_number,
                    'stimulus_type': stimulus_type,
                    'trial_type': trial_type,
                    'soa_condition_idx': soa_idx,
                    'soa_value_ms': soa_value_ms,  # Store exact SOA value in ms
                    'jitter_ms': current_jitter,
                    'timestamp_original': timestamp_original,
                    'timestamp_after_jitter': timestamp_after_jitter,
                    'timestamp_with_soa': timestamp_with_soa,  # New column with jitter + SOA
                    'base_ms': base_ms_raw,
                    'jittered_ms': jittered_ms_raw,
                    'soa_ms': soa_ms_raw
                })
        
        # Convert to DataFrame
        design_df = pd.DataFrame(design)
        
        # Shuffle within each trial type to maintain balance
        shuffled_design = []
        for trial_type in self.trial_counts.keys():
            type_subset = design_df[design_df['trial_type'] == trial_type].copy()
            type_subset = type_subset.sample(frac=1, random_state=participant_id).reset_index(drop=True)
            shuffled_design.append(type_subset)
        
        # Recombine, sort by timestamp, and reset index
        design_df = pd.concat(shuffled_design).reset_index(drop=True)
        design_df = design_df.sort_values(by='timestamp_after_jitter').reset_index(drop=True)
        
        return design_df
    
    def save_designs(self, designs):
        """
        Save generated designs to CSV files.
        
        Args:
            designs: List of participant design DataFrames
        """
        for design in designs:
            participant_id = design['participant_id'].iloc[0]
            filename = f'participant_{participant_id}_design.csv'
            filepath = os.path.join(self.output_dir, filename)
            design.to_csv(filepath, index=False)
            print(f"Saved design for participant {participant_id}")
    
    def validate_designs(self, designs):
        """
        Comprehensively validate the generated designs to ensure balance using Doegen-inspired
        validation checks and statistical methods.
        
        Args:
            designs: List of participant design DataFrames
        
        Returns:
            Dictionary with validation statistics
        """
        print("Starting design validation...")
        
        validation_results = {
            'stimulus_type': {},
            'trial_type': {},
            'soa_condition_idx': {},
            'jitter': {},
            'doegen_validation': {}
        }
        
        try:
            # Combine all designs for overall validation
            all_designs = pd.concat(designs)
            
            # ------- Basic Distribution Validation -------
            # Validate each factor by grouping by trial type
            for factor in ['stimulus_type', 'trial_type', 'soa_condition_idx']:
                grouped_validation = all_designs.groupby(['trial_type', factor]).size() / len(designs)
                validation_results[factor] = grouped_validation.to_dict()
            
            print("Basic distribution validation complete.")
            
            # ------- Doegen-inspired Validation Checks -------
            # 1. Check for First-Order Counterbalancing (transitions between conditions)
            print("Checking first-order counterbalancing...")
            transition_counts = {}
            for factor in ['stimulus_type', 'soa_condition_idx']:
                try:
                    # Count transitions across all designs
                    transitions = {}
                    for design in designs:
                        # Sort by timestamp to ensure we check transitions in presentation order
                        sorted_design = design.sort_values('timestamp_after_jitter')
                        for i in range(len(sorted_design) - 1):
                            curr = sorted_design.iloc[i][factor]
                            next_val = sorted_design.iloc[i+1][factor]
                            transition_key = f"{curr}->{next_val}"
                            if transition_key not in transitions:
                                transitions[transition_key] = 0
                            transitions[transition_key] += 1
                    
                    # Store transition counts
                    transition_counts[factor] = transitions
                    
                    # Calculate transition balance score (coefficient of variation)
                    values = np.array(list(transitions.values()))
                    cv = np.std(values) / np.mean(values) if np.mean(values) > 0 else 0
                    validation_results['doegen_validation'][f'{factor}_transition_cv'] = cv
                except Exception as e:
                    print(f"Warning: Error in counterbalancing check for {factor}: {str(e)}")
                    validation_results['doegen_validation'][f'{factor}_transition_cv'] = "Error"
            
            # 2. Check for Factor Orthogonality (independence between factors)
            print("Checking factor orthogonality...")
            orthogonality_results = {}
            for f1, f2 in combinations(['stimulus_type', 'trial_type', 'soa_condition_idx'], 2):
                try:
                    # Create contingency table
                    cont_table = pd.crosstab(all_designs[f1], all_designs[f2])
                    
                    # Calculate chi-square statistic for independence
                    chi2 = 0
                    total = cont_table.sum().sum()
                    for i in cont_table.index:
                        for j in cont_table.columns:
                            observed = cont_table.loc[i, j]
                            expected = cont_table.loc[i].sum() * cont_table[j].sum() / total
                            chi2 += ((observed - expected) ** 2) / expected if expected > 0 else 0
                    
                    orthogonality_results[f"{f1} vs {f2}_chi2"] = chi2
                except Exception as e:
                    print(f"Warning: Error in orthogonality check for {f1} vs {f2}: {str(e)}")
                    orthogonality_results[f"{f1} vs {f2}_chi2"] = "Error"
            
            validation_results['doegen_validation']['orthogonality'] = orthogonality_results
            
            # 3. Sequence Randomness Test (runs test for randomness)
            print("Checking sequence randomness...")
            randomness_results = {}
            for factor in ['stimulus_type', 'soa_condition_idx']:
                for design_idx, design in enumerate(designs[:5]):  # Test a sample of designs
                    try:
                        # Sort by timestamp
                        sorted_design = design.sort_values('timestamp_after_jitter')
                        sequence = sorted_design[factor].astype(str).tolist()
                        
                        # Count runs (consecutive identical values)
                        runs = 1
                        for i in range(1, len(sequence)):
                            if sequence[i] != sequence[i-1]:
                                runs += 1
                        
                        # Expected runs and std dev for a random sequence
                        n1 = len([x for x in sequence if x == sequence[0]])
                        n2 = len(sequence) - n1
                        if n1 > 0 and n2 > 0:  # Avoid division by zero
                            expected_runs = 1 + (2 * n1 * n2) / (n1 + n2)
                            std_runs = np.sqrt((2 * n1 * n2 * (2 * n1 * n2 - n1 - n2)) / 
                                            ((n1 + n2)**2 * (n1 + n2 - 1)))
                            
                            # Z-score for runs test
                            z_score = (runs - expected_runs) / std_runs if std_runs > 0 else 0
                            randomness_results[f"{factor}_design{design_idx}_z"] = z_score
                    except Exception as e:
                        print(f"Warning: Error in randomness check for {factor}, design {design_idx}: {str(e)}")
                        randomness_results[f"{factor}_design{design_idx}_z"] = "Error"
            
            validation_results['doegen_validation']['sequence_randomness'] = randomness_results
            
            # 4. Factor Sequence Autocorrelation Analysis
            print("Checking sequence autocorrelation...")
            autocorr_results = {}
            for factor in ['soa_condition_idx']:
                for design_idx, design in enumerate(designs[:5]):  # Test a sample of designs
                    try:
                        # Sort by timestamp
                        sorted_design = design.sort_values('timestamp_after_jitter')
                        
                        # Convert categorical variables to numeric for autocorrelation
                        if factor == 'stimulus_type':
                            # Map stimulus types to numbers
                            mapping = {stim: i for i, stim in enumerate(sorted_design[factor].unique())}
                            sequence = sorted_design[factor].map(mapping).tolist()
                        else:
                            sequence = sorted_design[factor].tolist()
                        
                        # Calculate lag-1 autocorrelation
                        mean_val = np.mean(sequence)
                        numerator = sum([(sequence[i] - mean_val) * (sequence[i+1] - mean_val) 
                                        for i in range(len(sequence)-1)])
                        denominator = sum([(val - mean_val)**2 for val in sequence])
                        
                        autocorr = numerator / denominator if denominator > 0 else 0
                        autocorr_results[f"{factor}_design{design_idx}_autocorr"] = autocorr
                    except Exception as e:
                        print(f"Warning: Error in autocorrelation check for {factor}, design {design_idx}: {str(e)}")
                        autocorr_results[f"{factor}_design{design_idx}_autocorr"] = "Error"
            
            validation_results['doegen_validation']['autocorrelation'] = autocorr_results
            
            # Plot factor distributions for visual validation
            print("Creating validation plots...")
            try:
                self._plot_factor_distributions(all_designs)
            except Exception as e:
                print(f"Warning: Error creating validation plots: {str(e)}")
            
            # Save detailed validation report to JSON
            validation_report_path = os.path.join(self.output_dir, 'validation_report.json')
            try:
                with open(validation_report_path, 'w') as f:
                    json.dump(validation_results, f, indent=4, default=str)
                print(f"Validation report saved to: {validation_report_path}")
            except Exception as e:
                print(f"Warning: Error saving validation report: {str(e)}")
            
            # Print summary of validation results
            print("\nValidation Results:")
            for factor, results in validation_results.items():
                if factor != 'doegen_validation':
                    print(f"\n{factor.replace('_', ' ').title()} Distribution:")
                    for key, value in sorted(results.items()):
                        print(f"  {key}: {value}")
            
            # Print Doegen validation summary
            print("\nDoegen Validation Checks:")
            for check_type, value in validation_results['doegen_validation'].items():
                if isinstance(value, dict):
                    print(f"  {check_type.replace('_', ' ').title()}:")
                    for subkey, subval in list(value.items())[:5]:  # Show first 5 items
                        print(f"    {subkey}: {subval}")
                    if len(value) > 5:
                        print(f"    ... and {len(value)-5} more (see validation_report.json)")
                else:
                    print(f"  {check_type.replace('_', ' ').title()}: {value}")
            
        except Exception as e:
            print(f"Error during validation: {str(e)}")
            print("Validation could not be completed. Returning partial results.")
        
        print("Validation complete.")
        return validation_results
    
    def _plot_factor_distributions(self, all_designs):
        """
        Create comprehensive visualizations of factor distributions for quality control,
        including Doegen-inspired validation plots.
        
        Args:
            all_designs: Combined DataFrame of all participant designs
        """
        plot_dir = os.path.join(self.output_dir, 'validation_plots')
        os.makedirs(plot_dir, exist_ok=True)
        
        # ------- Basic Distribution Plots -------
        # Create bar charts for factor distributions
        factors_to_plot = ['stimulus_type', 'soa_condition_idx']
        for factor in factors_to_plot:
            plt.figure(figsize=(10, 6))
            grouped = all_designs.groupby(['trial_type', factor]).size().unstack(fill_value=0)
            grouped.plot(kind='bar', stacked=True)
            plt.title(f'Distribution of {factor} Across Trial Types')
            plt.xlabel('Trial Type')
            plt.ylabel('Number of Occurrences')
            plt.legend(title=factor, bbox_to_anchor=(1.05, 1), loc='upper left')
            plt.tight_layout()
            plot_path = os.path.join(plot_dir, f'{factor}_distribution.png')
            plt.savefig(plot_path)
            plt.close()
        
        # Create heatmap for stimulus type distribution
        plt.figure(figsize=(12, 8))
        factor_pivot = all_designs.pivot_table(
            index='trial_type', 
            columns='stimulus_type', 
            aggfunc='size', 
            fill_value=0
        )
        sns.heatmap(factor_pivot, annot=True, cmap='YlGnBu', fmt='g')
        plt.title('Stimulus Type Distribution Across Trial Types')
        plt.tight_layout()
        heatmap_path = os.path.join(plot_dir, 'stimulus_type_heatmap.png')
        plt.savefig(heatmap_path)
        plt.close()
        
        # ------- Doegen-inspired Validation Plots -------
        
        # 1. Factor pairwise density plots (check for factor independence)
        factor_pairs = list(combinations(['stimulus_type', 'trial_type', 'soa_condition_idx'], 2))
        for f1, f2 in factor_pairs:
            plt.figure(figsize=(10, 8))
            
            # For categorical data, create a contingency table instead of 2D histogram
            contingency = pd.crosstab(all_designs[f1], all_designs[f2])
            
            # Plot as a heatmap
            sns.heatmap(contingency, annot=True, fmt='d', cmap='Blues')
            
            plt.xlabel(f1.replace('_', ' ').title())
            plt.ylabel(f2.replace('_', ' ').title())
            plt.title(f'Factor Independence Check: {f1} vs {f2}')
            plt.tight_layout()
            
            plot_path = os.path.join(plot_dir, f'factor_independence_{f1}_vs_{f2}.png')
            plt.savefig(plot_path)
            plt.close()
        
        # 2. Transition matrix heatmaps (first-order counterbalancing)
        for factor in ['stimulus_type', 'soa_condition_idx']:
            try:
                # Get unique values in sorted order for consistent matrix
                unique_vals = sorted(all_designs[factor].unique())
                n_vals = len(unique_vals)
                
                # Create transition matrix
                transitions = np.zeros((n_vals, n_vals))
                
                # Fill with transition counts from a sample of designs
                sample_designs = all_designs.groupby('participant_id').head(100)  # Limit sample size
                
                for pid, group in sample_designs.groupby('participant_id'):
                    sorted_group = group.sort_values('timestamp_after_jitter')
                    
                    # Map values to indices
                    if factor == 'stimulus_type':
                        mapping = {val: i for i, val in enumerate(unique_vals)}
                        sequence = sorted_group[factor].map(mapping).tolist()
                    else:
                        sequence = [list(unique_vals).index(val) for val in sorted_group[factor]]
                    
                    # Count transitions
                    for i in range(len(sequence) - 1):
                        from_idx = sequence[i]
                        to_idx = sequence[i + 1]
                        transitions[from_idx, to_idx] += 1
                
                # Plot transition matrix
                plt.figure(figsize=(10, 8))
                sns.heatmap(transitions, annot=True, fmt='.0f', cmap='YlGnBu',
                           xticklabels=unique_vals, yticklabels=unique_vals)
                plt.title(f'Transition Matrix for {factor}')
                plt.xlabel('To')
                plt.ylabel('From')
                plt.tight_layout()
                
                plot_path = os.path.join(plot_dir, f'transition_matrix_{factor}.png')
                plt.savefig(plot_path)
                plt.close()
            except Exception as e:
                print(f"Warning: Could not create transition matrix for {factor}: {str(e)}")
                continue
        
        # 3. Factor autocorrelation plots
        for factor in ['soa_condition_idx']:  # Only numeric factors
            try:
                # Sample a few designs
                autocorrs = []
                lags = list(range(1, 11))  # Lags 1 to 10
                
                for design_idx, (_, design) in enumerate(all_designs.groupby('participant_id')):
                    if design_idx >= 5:  # Limit to 5 participants
                        break
                        
                    sorted_design = design.sort_values('timestamp_after_jitter')
                    sequence = sorted_design[factor].astype(float).tolist()
                    
                    # Skip if too few data points
                    if len(sequence) < 20:
                        continue
                    
                    # Calculate autocorrelations for different lags
                    design_autocorrs = []
                    for lag in lags:
                        if lag >= len(sequence):
                            design_autocorrs.append(np.nan)
                            continue
                            
                        mean_val = np.mean(sequence)
                        numerator = sum([(sequence[i] - mean_val) * (sequence[i+lag] - mean_val) 
                                        for i in range(len(sequence)-lag)])
                        denominator = sum([(val - mean_val)**2 for val in sequence])
                        
                        autocorr = numerator / denominator if denominator > 0 else 0
                        design_autocorrs.append(autocorr)
                    
                    autocorrs.append(design_autocorrs)
                
                # Skip if no autocorrelations were calculated
                if not autocorrs:
                    continue
                    
                # Plot autocorrelations
                plt.figure(figsize=(10, 6))
                for i, autocorr in enumerate(autocorrs):
                    plt.plot(lags, autocorr, marker='o', label=f'Participant {i}')
                
                plt.axhline(y=0, color='r', linestyle='--', alpha=0.3)
                plt.xlabel('Lag')
                plt.ylabel('Autocorrelation')
                plt.title(f'Sequence Autocorrelation for {factor}')
                plt.legend()
                plt.grid(True, alpha=0.3)
                plt.tight_layout

                plt.tight_layout()
                
                plot_path = os.path.join(plot_dir, f'autocorrelation_{factor}.png')
                plt.savefig(plot_path)
                plt.close()
            except Exception as e:
                print(f"Warning: Could not create autocorrelation plot for {factor}: {str(e)}")
                continue
        
        # 4. SOA balance check across trial types
        try:
            plt.figure(figsize=(12, 6))
            
            # Convert soa_value_ms to string to avoid issues with pandas pivot
            all_designs['soa_value_ms_str'] = all_designs['soa_value_ms'].astype(str)
            
            soa_balance = pd.crosstab(
                index=all_designs['trial_type'],
                columns=all_designs['soa_value_ms_str'],
                margins=False
            )
            
            # Plot SOA balance
            sns.heatmap(soa_balance, annot=True, fmt='g', cmap='YlGnBu')
            plt.title('SOA Values Across Trial Types')
            plt.xlabel('SOA Value (ms)')
            plt.ylabel('Trial Type')
            plt.tight_layout()
            
            plot_path = os.path.join(plot_dir, 'soa_balance.png')
            plt.savefig(plot_path)
            plt.close()
        except Exception as e:
            print(f"Warning: Could not create SOA balance plot: {str(e)}")
            
        print(f"\nValidation plots saved to: {plot_dir}")


# ======================================================================
# AUDIO GENERATOR
# ======================================================================

class ParticipantAudioGenerator:
    """
    Generates TWO participant-specific audio files with embedded stimuli:
    1. Looming audio: Box breathing with directional looming sounds
    2. Tactile audio: Synchronized tactile stimuli with SOA offsets
    
    This class takes the counterbalanced designs and injects stimuli
    into audio files, creating unique pairs of audio for each participant
    with precisely timed stimuli.
    """
    
    def __init__(self, base_audio_path, output_dir, looming_stimuli_paths, tactile_stimulus_path, timestamps_csv_path=None):
        """
        Initialize the audio generator.
        
        Args:
            base_audio_path: Path to the base box breathing audio
            output_dir: Directory to save participant-specific audio files
            looming_stimuli_paths: Dictionary of paths to looming stimuli sounds
            tactile_stimulus_path: Path to the tactile stimulus sound
            timestamps_csv_path: Optional path to CSV with breathing timestamps (not needed when using design CSV)
        """
        # Store stimulus paths
        self.looming_stimuli = looming_stimuli_paths
        self.tactile_stimulus_path = tactile_stimulus_path
        
        # Load base audio (breathing exercise)
        self.base_audio_data, self.sample_rate = sf.read(base_audio_path)
        
        # Load tactile stimulus
        self.tactile_data, _ = sf.read(tactile_stimulus_path)
        
        # Load timestamps if provided (not needed when using design CSV directly)
        self.timestamps_df = None
        if timestamps_csv_path:
            self.timestamps_df = pd.read_csv(timestamps_csv_path)
        
        # Output directory for audio files
        self.output_dir = output_dir
        os.makedirs(output_dir, exist_ok=True)
    
    def generate_participant_audio(self, participant_id, design_df):
        """
        Generate two synchronized audio files for a participant:
        1. Looming stimuli embedded in box breathing audio (stereo)
        2. Tactile stimuli with SOA adjustments (mono content in stereo format)
        
        Uses precise millisecond values from design_df to ensure exact synchronization.
        
        Args:
            participant_id: Participant identifier
            design_df: DataFrame with stimuli design (MUST contain base_ms, jittered_ms, and soa_ms columns)
                
        Returns:
            Tuple (looming_audio, tactile_audio, injection_log)
        """
        # Create two separate audio streams
        looming_audio = self.base_audio_data.copy()  # Box breathing with looming sounds (stereo)
        tactile_audio = np.zeros_like(self.base_audio_data)  # Empty stereo track matching base audio shape
        
        # Ensure tactile stimulus is in stereo format (same content in both channels)
        # This converts mono to stereo if needed
        if self.tactile_data.ndim == 1:  # If tactile data is mono
            tactile_stereo = np.column_stack((self.tactile_data, self.tactile_data))
        else:
            # If already stereo, use as is
            tactile_stereo = self.tactile_data
            
        # Initialize log to track all stimulus injections
        injection_log = []
        
        # Verify that design_df contains required columns
        required_columns = ['stimulus_type', 'jittered_ms', 'soa_ms', 'soa_value_ms']
        missing_columns = [col for col in required_columns if col not in design_df.columns]
        if missing_columns:
            raise ValueError(f"Design DataFrame missing required columns: {missing_columns}. Cannot ensure precise synchronization.")
        
        # Process each trial in the design
        for _, row in design_df.iterrows():
            # Get exact millisecond timestamps from the design CSV
            jittered_ms = row['jittered_ms']  # For looming stimuli
            soa_ms = row['soa_ms']  # For tactile stimuli
            
            # Convert to sample indices
            looming_injection_point = int(jittered_ms / 1000 * self.sample_rate)
            tactile_injection_point = int(soa_ms / 1000 * self.sample_rate)
            
            # Get SOA value for logging
            soa_value_ms = row['soa_value_ms']
            
            # Load the appropriate looming stimulus based on design
            stim_type = row['stimulus_type']
            looming_path = self.looming_stimuli.get(stim_type)
            if looming_path is None:
                continue  # Skip if stimulus type not found
            
            # Read the looming stimulus audio file
            looming_data, _ = sf.read(looming_path)
            
            # Check if both stimuli fit within the audio bounds
            looming_fits = (looming_injection_point + len(looming_data) <= len(looming_audio))
            tactile_fits = (tactile_injection_point >= 0 and 
                           tactile_injection_point + len(tactile_stereo) <= len(tactile_audio))
            
            if looming_fits and tactile_fits:
                # Inject looming stimulus into the first audio track
                looming_audio[looming_injection_point:looming_injection_point+len(looming_data)] += looming_data
                
                # Inject tactile stimulus into the second audio track (now using stereo format)
                tactile_audio[tactile_injection_point:tactile_injection_point+len(tactile_stereo)] += tactile_stereo
                
                # Log details of this injection for record-keeping
                injection_log.append({
                    'participant_id': participant_id,
                    'trial_number': row.get('trial_number', 0),
                    'stimulus_type': stim_type,
                    'soa_condition_idx': row.get('soa_condition_idx', 0),
                    'soa_value_ms': soa_value_ms,
                    'looming_injection_sample': looming_injection_point,
                    'tactile_injection_sample': tactile_injection_point,
                    'looming_time_ms': looming_injection_point / self.sample_rate * 1000,
                    'tactile_time_ms': tactile_injection_point / self.sample_rate * 1000,
                    'jittered_ms': jittered_ms,
                    'soa_ms': soa_ms
                })
        
        return looming_audio, tactile_audio, pd.DataFrame(injection_log)
    
    def save_participant_files(self, participant_id, looming_audio, tactile_audio, injection_log, design_filename):
        """
        Save participant-specific audio and log files.
        
        Args:
            participant_id: Participant identifier
            looming_audio: Modified audio with looming stimuli
            tactile_audio: Audio track with tactile stimuli
            injection_log: DataFrame with injection details
            design_filename: Base design filename for naming consistency
        
        Returns:
            Tuple of saved file paths
        """
        # Use the design CSV base name (remove .csv extension)
        base_name = os.path.splitext(design_filename)[0]
        
        # Define filenames for output files
        looming_filename = f"{base_name}_looming.wav"
        tactile_filename = f"{base_name}_tactile.wav"
        log_filename = f"{base_name}_stimuli_log.csv"
        
        # Create full paths
        looming_path = os.path.join(self.output_dir, looming_filename)
        tactile_path = os.path.join(self.output_dir, tactile_filename)
        log_path = os.path.join(self.output_dir, log_filename)
        
        # Save the audio files
        sf.write(looming_path, looming_audio, self.sample_rate)
        sf.write(tactile_path, tactile_audio, self.sample_rate)
        
        # Save the detailed injection log (for analysis)
        injection_log.to_csv(log_path, index=False)
        
        print(f"Looming audio saved: {looming_filename}")
        print(f"Tactile audio saved: {tactile_filename}")
        print(f"Injection log saved: {log_filename}")
        
        return looming_path, tactile_path, log_path


# ======================================================================
# MAIN FUNCTION
# ======================================================================

def main():
    """
    Main function to run the complete PPS experiment setup process:
    1. Generate counterbalanced designs for multiple participants
    2. Create two synchronized audio files per participant:
       - Looming stimuli embedded in box breathing audio
       - Tactile stimuli with SOA adjustments
    
    The script creates and saves:
    1. Design CSV files in the DESIGN_OUTPUT_DIR directory, including:
       - exact SOA values in milliseconds
       - precise timestamps with jitter and SOA adjustments
    2. Two WAV files per participant in the AUDIO_OUTPUT_DIR directory:
       - participant_X_design_looming.wav (box breathing + looming stimuli)
       - participant_X_design_tactile.wav (synchronized tactile stimuli)
    3. Log files with detailed timing information
    
    CRITICAL: This implementation ensures precise synchronization between the audio files
    by directly referencing timestamps from the design CSV file rather than recalculating them.
    """
    # ===== Ensure output directories exist =====
    os.makedirs(DESIGN_OUTPUT_DIR, exist_ok=True)
    os.makedirs(AUDIO_OUTPUT_DIR, exist_ok=True)
    
    # ===== Step 1: Generate counterbalanced designs with exact SOA values =====
    print("===== STEP 1: GENERATING COUNTERBALANCED DESIGNS =====")
    print(f"Using SOA conditions (ms): {SOA_CONDITIONS_MS}")
    print("Each design CSV will contain exact SOA values and precise timestamps")
    
    counterbalancer = ComprehensiveCounterbalancer(
        output_dir=DESIGN_OUTPUT_DIR,
        soa_conditions=SOA_CONDITIONS_MS,
        trial_counts=TRIAL_COUNTS,
        jitter_options=JITTER_OPTIONS_MS
    )
    
    designs = counterbalancer.generate_comprehensive_design(
        num_participants=NUM_PARTICIPANTS, 
        reference_csv_path=TIMESTAMPS_CSV_PATH
    )
    
    # Save designs to CSV files - these now include exact SOA values and timestamp_with_soa
    counterbalancer.save_designs(designs)
    
    # Run comprehensive Doegen validation checks
    print("\n===== RUNNING DOEGEN COUNTERBALANCING VALIDATION CHECKS =====")
    print("Performing comprehensive validation including:")
    print("  - First-order counterbalancing (transition matrices)")
    print("  - Factor orthogonality/independence")
    print("  - Sequence randomness (runs test)")
    print("  - Autocorrelation analysis")
    print("  - SOA balance verification")
    
    try:
        validation_results = counterbalancer.validate_designs(designs)
        
        print("\n===== DOEGEN VALIDATION SUMMARY =====")
        if 'doegen_validation' in validation_results:
            # Check transition balance
            if 'stimulus_type_transition_cv' in validation_results['doegen_validation']:
                cv = validation_results['doegen_validation']['stimulus_type_transition_cv']
                if not isinstance(cv, str) and cv < 0.2:
                    print("✓ Stimulus transitions are well-balanced (CV < 0.2)")
                else:
                    print(f"⚠ Stimulus transitions show some imbalance (CV = {cv})")
                    
            # Check orthogonality
            if 'orthogonality' in validation_results['doegen_validation']:
                ortho = validation_results['doegen_validation']['orthogonality']
                non_error_values = [v for v in ortho.values() if not isinstance(v, str)]
                if non_error_values:
                    mean_chi2 = np.mean(non_error_values)
                    if mean_chi2 < 20:
                        print("✓ Factors are sufficiently orthogonal/independent")
                    else:
                        print(f"⚠ Some factors may have dependencies (mean χ² = {mean_chi2:.2f})")
                    
            # Check sequence randomness
            if 'sequence_randomness' in validation_results['doegen_validation']:
                random_vals = [v for v in validation_results['doegen_validation']['sequence_randomness'].values() 
                              if not isinstance(v, str)]
                if random_vals:
                    mean_abs_z = np.mean([abs(z) for z in random_vals])
                    if mean_abs_z < 1.96:
                        print("✓ Sequences appear sufficiently random (|z| < 1.96)")
                    else:
                        print(f"⚠ Sequences may have randomness issues (mean |z| = {mean_abs_z:.2f})")
                    
            # Check autocorrelation
            if 'autocorrelation' in validation_results['doegen_validation']:
                autocorr_vals = [v for v in validation_results['doegen_validation']['autocorrelation'].values()
                                if not isinstance(v, str)]
                if autocorr_vals:
                    mean_autocorr = np.mean([abs(a) for a in autocorr_vals])
                    if mean_autocorr < 0.2:
                        print("✓ Low sequential autocorrelation (< 0.2)")
                    else:
                        print(f"⚠ Some sequential dependencies detected (mean |autocorr| = {mean_autocorr:.2f})")
    
    except Exception as e:
        print(f"Warning: Validation checks encountered an error: {str(e)}")
        print("Continuing with design generation...")
    
    print("\nSee validation_plots directory for detailed visual analyses")
    
    # ===== Step 2: Generate participant-specific audio files =====
    print("\n===== STEP 2: GENERATING PARTICIPANT AUDIO FILES =====")
    print(f"Creating TWO audio files per participant:")
    print(f"  - Looming audio: Box breathing with directional looming sounds")
    print(f"  - Tactile audio: Synchronized tactile stimuli with SOA offsets")
    print(f"CRITICAL: Using exact millisecond timestamps from design CSV for precise synchronization")
    
    # Initialize the audio generator, now with timestamps_csv_path as optional
    audio_generator = ParticipantAudioGenerator(
        base_audio_path=BASE_AUDIO_PATH,
        output_dir=AUDIO_OUTPUT_DIR,
        looming_stimuli_paths=LOOMING_STIMULI,
        tactile_stimulus_path=TACTILE_STIMULUS_PATH,
        timestamps_csv_path=TIMESTAMPS_CSV_PATH  # Now optional
    )
    
    # Process each design and generate audio pairs
    for i, design_df in enumerate(designs):
        participant_id = design_df['participant_id'].iloc[0]
        design_filename = f"participant_{participant_id}_design.csv"
        
        print(f"\nProcessing participant {participant_id} ({i+1}/{NUM_PARTICIPANTS})...")
        
        # Verify that design_df contains necessary timing columns
        if 'jittered_ms' not in design_df.columns or 'soa_ms' not in design_df.columns:
            print(f"Warning: Design for participant {participant_id} missing critical timing columns.")
            print("Reading design from saved CSV file to ensure correct timing...")
            
            # Re-read the design CSV to ensure we have the most accurate timing info
            design_file_path = os.path.join(DESIGN_OUTPUT_DIR, design_filename)
            if os.path.exists(design_file_path):
                design_df = pd.read_csv(design_file_path)
                print("Successfully loaded design from CSV with all timing information.")
            else:
                print(f"Error: Cannot find design file {design_file_path}")
                continue
        
        # Generate two synchronized audio tracks for this participant
        try:
            looming_audio, tactile_audio, injection_log = audio_generator.generate_participant_audio(
                participant_id, 
                design_df
            )
            
            # Save both audio files and the log
            audio_generator.save_participant_files(
                participant_id, 
                looming_audio, 
                tactile_audio, 
                injection_log, 
                design_filename
            )
            
            # Verify synchronization
            print("Verifying synchronization...")
            if len(injection_log) > 0:
                for i, row in injection_log.head(3).iterrows():
                    looming_time = row['looming_time_ms']
                    tactile_time = row['tactile_time_ms']
                    expected_diff = row['soa_value_ms']
                    actual_diff = tactile_time - looming_time
                    
                    print(f"  Trial {i+1}: SOA expected={expected_diff}ms, actual={actual_diff:.2f}ms, " +
                          f"difference={abs(actual_diff-expected_diff):.4f}ms")
                
                if len(injection_log) > 3:
                    print(f"  ... and {len(injection_log) - 3} more trials")
            else:
                print("  Warning: No stimuli were injected - check design parameters")
                
        except Exception as e:
            print(f"Error generating audio for participant {participant_id}: {str(e)}")
            continue
    
    print(f"\nSUMMARY:")
    print(f"All designs generated and saved to: {DESIGN_OUTPUT_DIR}")
    print(f"All audio files generated and saved to: {AUDIO_OUTPUT_DIR}")
    print(f"Created {NUM_PARTICIPANTS} participant designs with {len(SOA_CONDITIONS_MS)} SOA conditions")
    print(f"Each participant has TWO audio files: one with looming stimuli and one with tactile stimuli")
    print(f"Exact SOA values used (ms): {SOA_CONDITIONS_MS}")
    print(f"Every design CSV contains precise millisecond timing for exact synchronization")
    print(f"Process complete.")

# ===== For Jupyter notebook use =====
# This cell can be run directly in a Jupyter notebook
if __name__ == "__main__":
    main()
```

# PPS Practice Stimulus Generator - Concise Outline

This code generates short practice stimuli for a Peripersonal Space (PPS) experiment with these key features:

### What it Creates:
- 5 different randomized versions of practice stimuli
- Each version includes two synchronized audio files:
  1. A file containing looming sounds (approaching from left/front/right)
  2. A file containing tactile stimuli (vibrations)
- CSV log files documenting precise stimulus timing

### Content Structure:
- Each version contains 12 trials with 4-second intervals between them:
  - 4 baseline trials (tactile vibration only)
  - 4 catch trials (looming sound only) 
  - 4 PPS trials (both looming and tactile with random SOA)
- Different randomized order in each version

### Technical Approach:
- Creates short audio files (approximately 1-2 minutes)
- Maintains precise synchronization between paired audio files
- Uses random SOA (Stimulus Onset Asynchrony) for PPS trials
- Includes trial timing information in CSV logs for verification

The program outputs all files to the specified directory, ready for use in participant practice/training sessions.

```python
# ======================================================================
# PRACTICE STIMULI GENERATOR FOR PPS EXPERIMENT
# ======================================================================
# 
# This script generates 5 different versions of practice stimuli for the 
# Peripersonal Space (PPS) experiment. Each version contains:
#
# 1. 12 trials divided into:
#    - 4 baseline trials (tactile stimuli only)
#    - 4 catch trials (looming stimuli only)
#    - 4 PPS trials (both looming and tactile with random SOA)
#
# 2. Two synchronized audio files:
#    - One for looming stimuli
#    - One for tactile stimuli
#
# Each of the 5 versions has a different ordering of these trials.
#
# PRACTICE STIMULI PURPOSE:
# These practice files will help familiarize participants with the different 
# trial types before the actual experiment starts.

import os
import numpy as np
import pandas as pd
import soundfile as sf
import matplotlib.pyplot as plt
from datetime import datetime
import random
import warnings

# Suppress warnings for cleaner output
warnings.filterwarnings("ignore")

# ======================================================================
# USER-CONFIGURABLE PARAMETERS
# ======================================================================

# File paths - UPDATE THESE TO MATCH YOUR SYSTEM
BASE_DIR = r"C:\Users\cogpsy-vrlab\Documents\PPS_module\BreathingPilot"

# Box breathing audio
BASE_AUDIO_PATH = os.path.join(BASE_DIR, "AudioInstructionsBoxBreathing", "TargetStimuli", 
                              "6. Box Breathing Combined (Segment Injected).wav")

# Timestamp reference for breathing phases
TIMESTAMPS_CSV_PATH = os.path.join(BASE_DIR, "AudioInstructionsBoxBreathing", "TargetStimuli", 
                                  "BoxBreathing_extended_timestamps.csv")

# Output directory for practice stimuli
PRACTICE_STIMULI_DIR = r"C:\Users\cogpsy-vrlab\Documents\PPS_module\BreathingPilot\PracticeStimuli"

# Looming stimuli (directional audio)
LOOMING_STIMULI = {
    'right': os.path.join(BASE_DIR, "PPS_Experiment_Module", "right_az90_FABIAN_HRIR_natural.wav"),
    'front': os.path.join(BASE_DIR, "PPS_Experiment_Module", "front_az0_FABIAN_HRIR_natural.wav"),
    'left': os.path.join(BASE_DIR, "PPS_Experiment_Module", "left_az-90_FABIAN_HRIR_natural.wav")
}

# Tactile stimulus (to be synchronized with looming stimuli + SOA)
TACTILE_STIMULUS_PATH = os.path.join(BASE_DIR, "PPS_Experiment_Module", "tactile_stimulus.wav")

# Range of SOA values (in milliseconds) for PPS trials
SOA_RANGE = [100, 1500]  # Random SOA between 100ms and 1500ms

# Jitter options in milliseconds (temporal variability)
JITTER_OPTIONS_MS = [100, 200, 300, 400, 500]

# Number of iterations (different versions)
NUM_ITERATIONS = 5

# ======================================================================
# PRACTICE STIMULI GENERATOR
# ======================================================================

class PracticeStimuli:
    """
    Generates practice stimuli for the PPS experiment with balanced trial types:
    - Baseline trials (tactile only)
    - Catch trials (looming only)
    - PPS trials (both looming and tactile with random SOA)
    """
    
    def __init__(self, output_dir, looming_stimuli_paths, tactile_stimulus_path, base_audio_path=None, timestamps_csv_path=None):
        """
        Initialize the practice stimuli generator
        
        Args:
            output_dir: Directory to save practice stimuli
            looming_stimuli_paths: Dictionary of paths to looming stimuli sounds
            tactile_stimulus_path: Path to the tactile stimulus sound
            base_audio_path: Path to an audio file to get sample rate (not used for content)
            timestamps_csv_path: Not used in this simplified version
        """
        # Create output directory
        self.output_dir = output_dir
        os.makedirs(output_dir, exist_ok=True)
        
        # Store stimulus paths
        self.looming_stimuli = looming_stimuli_paths
        self.tactile_stimulus_path = tactile_stimulus_path
        
        # Set sample rate (typically 44100 Hz)
        self.sample_rate = 44100
        
        # Check if any audio file exists to get sample rate
        if base_audio_path:
            try:
                _, loaded_sample_rate = sf.read(base_audio_path)
                self.sample_rate = loaded_sample_rate
                print(f"Using sample rate from audio file: {self.sample_rate} Hz")
            except Exception as e:
                print(f"Could not load audio for sample rate: {e}")
                print(f"Using default sample rate: {self.sample_rate} Hz")
        
        # Load tactile stimulus
        try:
            self.tactile_data, _ = sf.read(tactile_stimulus_path)
            print(f"Tactile stimulus loaded: {tactile_stimulus_path}")
        except Exception as e:
            print(f"Error loading tactile stimulus: {e}")
            raise
        
        # For storing looming stimulus data
        self.loaded_looming_stimuli = {}
        for stim_type, path in self.looming_stimuli.items():
            try:
                self.loaded_looming_stimuli[stim_type], _ = sf.read(path)
                print(f"Looming stimulus loaded: {stim_type}")
            except Exception as e:
                print(f"Error loading looming stimulus {stim_type}: {e}")
    
    def generate_practice_iteration(self, iteration_num):
        """
        Generate a single practice iteration with 12 trials:
        - 4 baseline trials (tactile only)
        - 4 catch trials (looming only)
        - 4 PPS trials (looming + tactile with random SOA)
        
        Each trial is separated by only 4 seconds.
        
        Args:
            iteration_num: Iteration number (1-5)
        
        Returns:
            Tuple (looming_audio, tactile_audio, trial_log)
        """
        print(f"\nGenerating practice iteration {iteration_num}...")
        
        # Create trial types (4 of each type)
        trial_types = ['baseline'] * 4 + ['catch'] * 4 + ['pps'] * 4
        
        # Use a different random seed for each iteration to ensure different orderings
        np.random.seed(42 + iteration_num)
        np.random.shuffle(trial_types)
        
        # Create stimulus directions (balanced across types)
        stimulus_directions = ['left', 'front', 'right'] * 4
        np.random.shuffle(stimulus_directions)
        
        # Get the duration of the looming stimuli (use the first one as reference)
        first_looming_key = list(self.loaded_looming_stimuli.keys())[0]
        looming_duration_samples = len(self.loaded_looming_stimuli[first_looming_key])
        
        # Get the duration of the tactile stimulus
        if self.tactile_data.ndim == 1:  # If tactile data is mono
            tactile_stereo = np.column_stack((self.tactile_data, self.tactile_data))
        else:
            tactile_stereo = self.tactile_data
        tactile_duration_samples = len(tactile_stereo)
        
        # Calculate the total duration needed (in samples)
        # 4 seconds between trials + buffer at beginning and end
        interval_samples = 4 * self.sample_rate  # 4 seconds between trials
        max_stimulus_duration = max(looming_duration_samples, tactile_duration_samples)
        
        # Total duration: start buffer + (12 trials * (interval + max stimulus duration)) + end buffer
        total_samples = 2 * self.sample_rate + (12 * (interval_samples + max_stimulus_duration)) + 2 * self.sample_rate
        
        # Create two separate audio streams (initialize as silence)
        looming_audio = np.zeros((total_samples, 2))  # Stereo silence
        tactile_audio = np.zeros((total_samples, 2))  # Stereo silence
        
        # Track injection details
        injection_log = []
        
        # Process each trial
        for trial_idx, (trial_type, stim_direction) in enumerate(zip(trial_types, stimulus_directions)):
            # Calculate position for this trial (in samples)
            # Start with 2 second buffer, then add trial intervals
            position_samples = 2 * self.sample_rate + trial_idx * (interval_samples + max_stimulus_duration)
            position_ms = (position_samples / self.sample_rate) * 1000
            
            # Apply random jitter (100-500ms)
            jitter_ms = np.random.choice(JITTER_OPTIONS_MS)
            jittered_ms = position_ms + jitter_ms
            jittered_samples = int(jittered_ms / 1000 * self.sample_rate)
            
            # For PPS trials, generate a random SOA
            soa_ms = 0
            if trial_type == 'pps':
                soa_ms = np.random.randint(SOA_RANGE[0], SOA_RANGE[1])
            
            # Calculate injection points
            looming_injection_point = jittered_samples
            tactile_injection_point = jittered_samples + int((soa_ms / 1000) * self.sample_rate)
            
            # Perform the appropriate injections based on trial type
            if trial_type in ['catch', 'pps']:
                # Inject looming stimulus
                looming_data = self.loaded_looming_stimuli.get(stim_direction)
                if looming_data is not None and looming_injection_point + len(looming_data) <= len(looming_audio):
                    looming_audio[looming_injection_point:looming_injection_point+len(looming_data)] += looming_data
            
            if trial_type in ['baseline', 'pps']:
                # Inject tactile stimulus
                if tactile_injection_point + len(tactile_stereo) <= len(tactile_audio):
                    tactile_audio[tactile_injection_point:tactile_injection_point+len(tactile_stereo)] += tactile_stereo
            
            # Log trial details
            injection_log.append({
                'iteration_num': iteration_num,
                'trial_number': trial_idx + 1,
                'trial_type': trial_type,
                'stimulus_direction': stim_direction,
                'position_ms': position_ms,
                'jitter_ms': jitter_ms,
                'jittered_ms': jittered_ms,
                'soa_ms': soa_ms,
                'looming_injection_sample': looming_injection_point,
                'tactile_injection_sample': tactile_injection_point,
                'looming_time_seconds': looming_injection_point / self.sample_rate,
                'tactile_time_seconds': tactile_injection_point / self.sample_rate
            })
        
        return looming_audio, tactile_audio, pd.DataFrame(injection_log)
    
    def save_practice_files(self, iteration_num, looming_audio, tactile_audio, trial_log):
        """
        Save practice stimuli files for an iteration
        
        Args:
            iteration_num: Iteration number
            looming_audio: Modified audio with looming stimuli
            tactile_audio: Audio track with tactile stimuli
            trial_log: DataFrame with trial details
        
        Returns:
            Tuple of saved file paths
        """
        # Define filenames
        looming_filename = f"practice_iteration_{iteration_num}_looming.wav"
        tactile_filename = f"practice_iteration_{iteration_num}_tactile.wav"
        log_filename = f"practice_iteration_{iteration_num}_log.csv"
        
        # Create full paths
        looming_path = os.path.join(self.output_dir, looming_filename)
        tactile_path = os.path.join(self.output_dir, tactile_filename)
        log_path = os.path.join(self.output_dir, log_filename)
        
        # Save the audio files
        sf.write(looming_path, looming_audio, self.sample_rate)
        sf.write(tactile_path, tactile_audio, self.sample_rate)
        
        # Save the trial log
        trial_log.to_csv(log_path, index=False)
        
        print(f"Saved practice files for iteration {iteration_num}:")
        print(f"  Looming audio: {looming_filename}")
        print(f"  Tactile audio: {tactile_filename}")
        print(f"  Trial log: {log_filename}")
        
        return looming_path, tactile_path, log_path
    
    # Removed plotting function as requested


# ======================================================================
# MAIN FUNCTION
# ======================================================================

def main():
    """
    Main function to generate 5 different iterations of practice stimuli for the PPS experiment.
    
    Each iteration contains 12 trials (4 baseline, 4 catch, 4 PPS) with randomized order.
    For each iteration, two synchronized audio files are generated:
    - One for looming stimuli
    - One for tactile stimuli
    
    Trials are separated by 4 seconds for a much shorter practice session.
    """
    print("===== GENERATING PPS EXPERIMENT PRACTICE STIMULI =====")
    print(f"Creating {NUM_ITERATIONS} different iterations of practice stimuli")
    print(f"Each iteration will contain 12 trials with only 4 seconds between trials:")
    print(f"  - 4 baseline trials (tactile only)")
    print(f"  - 4 catch trials (looming only)")
    print(f"  - 4 PPS trials (both with random SOA)")
    print(f"Output directory: {PRACTICE_STIMULI_DIR}")
    
    # Initialize the practice stimuli generator
    practice_generator = PracticeStimuli(
        output_dir=PRACTICE_STIMULI_DIR,
        looming_stimuli_paths=LOOMING_STIMULI,
        tactile_stimulus_path=TACTILE_STIMULUS_PATH,
        base_audio_path=BASE_AUDIO_PATH,
        timestamps_csv_path=TIMESTAMPS_CSV_PATH
    )
    
    # Generate practice stimuli for each iteration
    for iteration in range(1, NUM_ITERATIONS + 1):
        try:
            # Generate the practice stimuli
            looming_audio, tactile_audio, trial_log = practice_generator.generate_practice_iteration(iteration)
            
            # Save the files
            practice_generator.save_practice_files(iteration, looming_audio, tactile_audio, trial_log)
            
            # Summary of this iteration
            print(f"\nIteration {iteration} summary:")
            trial_types = trial_log['trial_type'].value_counts()
            for trial_type, count in trial_types.items():
                print(f"  {trial_type}: {count} trials")
            
            # Show SOA distribution for PPS trials
            pps_trials = trial_log[trial_log['trial_type'] == 'pps']
            if not pps_trials.empty:
                min_soa = pps_trials['soa_ms'].min()
                max_soa = pps_trials['soa_ms'].max()
                mean_soa = pps_trials['soa_ms'].mean()
                print(f"  PPS trials SOA range: {min_soa}-{max_soa}ms (mean: {mean_soa:.1f}ms)")
            
        except Exception as e:
            print(f"Error generating practice iteration {iteration}: {e}")
    
    print("\n===== PRACTICE STIMULI GENERATION COMPLETE =====")
    print(f"Created {NUM_ITERATIONS} practice iterations with 12 trials each")
    print(f"All files saved to: {PRACTICE_STIMULI_DIR}")
    print("Each iteration contains two audio files (looming and tactile) with:")
    print("  - 4 baseline trials (tactile stimulus only)")
    print("  - 4 catch trials (looming stimulus only)")
    print("  - 4 PPS trials (both looming and tactile with random SOA)")
    print("Each iteration has a different randomized order of trials")
    print("There are only 4 seconds between trials for shorter practice sessions")
    print("\nEach CSV file records the exact timing of when looming and tactile stimuli are played")

if __name__ == "__main__":
    main()
```

```python
"""
Dynamic Peripersonal Space (PPS) Experiment Generator

This script generates customizable PPS experiments with dynamic trial counts
and automatic adjustment of experiment length based on the number of trials.

The breathing pattern is fixed by the base audio track (box breathing),
but you can control which breathing phases to use for stimuli presentation,
how many trials per condition, and what percentage of trials are catch trials.
"""

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import soundfile as sf
import scipy.signal as signal
from datetime import datetime
from itertools import product, combinations
import warnings
import json
import math
import random

# ======================================================================
# CONFIGURABLE PARAMETERS - MODIFY THESE TO CUSTOMIZE YOUR EXPERIMENT
# ======================================================================

# Experiment structure
EXPERIMENT_CONFIG = {
    'repetitions_per_condition': 1,  # How many times each condition is repeated
    'included_conditions': {
        'inhalation': True,  # Include inhalation trials
        'exhalation': False,  # Include exhalation trials
        'baseline': False,    # Include baseline trials
    },
    'catch_trial_percentage': 10,  # Percentage of total trials that should be catch trials
    'debug_mode': False,    # Enable for verbose output and diagnostics
}

# SOA conditions in milliseconds (Stimulus Onset Asynchrony)
SOA_CONDITIONS_MS = [190, 400, 700, 1000, 1500]

# Jitter options in milliseconds (temporal variability)
JITTER_OPTIONS_MS = [100, 200, 300, 400, 500]

# File paths - UPDATE THESE TO MATCH YOUR SYSTEM
BASE_DIR = r"C:\Users\cogpsy-vrlab\Documents\GitHub\BreathingSpace"

# Original audio paths (update if needed)
ORIG_AUDIO_DIR = r"C:\Users\cogpsy-vrlab\Documents\PPS_module\BreathingPilot"

# Box breathing audio
BASE_AUDIO_PATH = os.path.join(ORIG_AUDIO_DIR, "AudioInstructionsBoxBreathing", "TargetStimuli", 
                              "6. Box Breathing Combined (Segment Injected).wav")

# Timestamp reference for breathing phases
TIMESTAMPS_CSV_PATH = os.path.join(ORIG_AUDIO_DIR, "AudioInstructionsBoxBreathing", "TargetStimuli", 
                                  "BoxBreathing_extended_timestamps.csv")

# Output directories
DESIGN_OUTPUT_DIR = os.path.join(BASE_DIR, "ExperimentLog")
AUDIO_OUTPUT_DIR = os.path.join(BASE_DIR, "ExperimentAudio")

# Looming stimuli (directional audio)
LOOMING_STIMULI = {
    'right': os.path.join(ORIG_AUDIO_DIR, "PPS_Experiment_Module", "right_az90_FABIAN_HRIR_natural.wav"),
    'front': os.path.join(ORIG_AUDIO_DIR, "PPS_Experiment_Module", "front_az0_FABIAN_HRIR_natural.wav"),
    'left': os.path.join(ORIG_AUDIO_DIR, "PPS_Experiment_Module", "left_az-90_FABIAN_HRIR_natural.wav")
}

# Tactile stimulus (to be synchronized with looming stimuli + SOA)
TACTILE_STIMULUS_PATH = os.path.join(ORIG_AUDIO_DIR, "PPS_Experiment_Module", "tactile_stimulus.wav")

# Number of participants to generate designs for
NUM_PARTICIPANTS = 100

# ======================================================================
# DYNAMIC TRIAL COUNT CALCULATOR
# ======================================================================

def calculate_trial_counts():
    """
    Calculate trial counts based on the experiment configuration.
    
    Returns:
        Dictionary with trial counts for each condition
    """
    # Get configuration
    reps = EXPERIMENT_CONFIG['repetitions_per_condition']
    selected_conditions = [cond for cond, include in EXPERIMENT_CONFIG['included_conditions'].items() if include]
    
    # Each condition has three stimulus types (left, front, right)
    stimuli_per_condition = 3
    
    # Calculate basic trial counts for selected conditions
    basic_trial_counts = {cond: reps * stimuli_per_condition for cond in selected_conditions}
    
    # Calculate total number of trials from selected conditions
    total_trials = sum(basic_trial_counts.values())
    
    # Calculate catch trials based on percentage
    catch_percentage = EXPERIMENT_CONFIG['catch_trial_percentage']
    total_with_catch = total_trials / (1 - catch_percentage/100)
    catch_trials = round(total_with_catch * (catch_percentage/100))
    
    # Add catch trials to the count
    final_counts = basic_trial_counts.copy()
    final_counts['catch'] = catch_trials
    
    if EXPERIMENT_CONFIG['debug_mode']:
        print(f"Trial counts calculated:")
        for cond, count in final_counts.items():
            print(f"  {cond}: {count} trials")
        print(f"  Total: {sum(final_counts.values())} trials")
    
    return final_counts

# ======================================================================
# COUNTERBALANCED DESIGN GENERATOR
# ======================================================================

class DynamicCounterbalancer:
    """
    Creates counterbalanced experimental designs with flexible trial counts
    based on configuration parameters.
    """
    
    def __init__(self, output_dir, soa_conditions, jitter_options):
        """
        Initialize the counterbalancer with experiment parameters.
        
        Args:
            output_dir: Directory to save design files
            soa_conditions: List of SOA values in milliseconds
            jitter_options: List of jitter values in milliseconds
        """
        # Ensure output directory exists
        self.output_dir = output_dir
        os.makedirs(output_dir, exist_ok=True)
        
        # Calculate trial counts based on configuration
        self.trial_counts = calculate_trial_counts()
        
        # Define experimental factors
        self.factors = {
            'stimulus_type': ['right', 'front', 'left'],  # Stimulus directions
            'trial_type': list(self.trial_counts.keys()),  # Active trial types
            'soa_condition': list(range(len(soa_conditions))),  # SOA condition indices
            'jitter': jitter_options  # Jitter in milliseconds
        }
        
        # Store actual SOA values for reference
        self.soa_values_ms = soa_conditions
        
        # Default timing parameters
        self.base_time = pd.Timestamp('2025-03-01 09:00:00')
        self.trial_duration = 8  # Average seconds between trials
        
        # This will be set if a reference CSV is provided
        self.reference_csv_path = None

    def generate_comprehensive_design(self, num_participants=100, reference_csv_path=None):
        """
        Generate a counterbalanced design for multiple participants.
        
        Args:
            num_participants: Number of participant designs to generate
            reference_csv_path: Path to CSV with breathing exercise timestamps
        
        Returns:
            List of DataFrames, one for each participant
        """
        self.reference_csv_path = reference_csv_path
        designs = []
        
        # Set a fixed seed for reproducibility
        np.random.seed(42)
        
        # Generate design for each participant
        for participant_id in range(num_participants):
            design = self._generate_single_participant_design(participant_id)
            designs.append(design)
        
        return designs
    
    def _generate_single_participant_design(self, participant_id):
        """
        Generate a balanced design for a single participant with precise timestamps.
        
        Args:
            participant_id: Unique identifier for the participant
        
        Returns:
            DataFrame with complete experimental design including timestamps
        """
        design = []
        
        # If a reference CSV is provided, use it for timing
        if self.reference_csv_path is not None:
            timestamp_df = pd.read_csv(self.reference_csv_path)
            csv_milliseconds = timestamp_df['milliseconds'].astype(int).values
            total_trials = sum(self.trial_counts.values())
            
            # Handle case where there aren't enough timestamps
            if total_trials > len(csv_milliseconds):
                if EXPERIMENT_CONFIG['debug_mode']:
                    print(f"Warning: Not enough timestamps in CSV. Needed {total_trials}, found {len(csv_milliseconds)}.")
                    print(f"Will recycle timestamps as needed.")
                
                # Create a repeated version of timestamps to cover all trials
                repeats_needed = math.ceil(total_trials / len(csv_milliseconds))
                csv_milliseconds = np.tile(csv_milliseconds, repeats_needed)[:total_trials]
            
            csv_index = 0  # Track which CSV row to use for each trial
        
        # Generate designs for each trial type (inhalation, exhalation, etc.)
        for trial_type, count in self.trial_counts.items():
            # Skip conditions with zero trials
            if count == 0:
                continue
                
            # Create a balanced stimulus sequence for each trial type
            stimulus_sequence = []
            for stim_type in self.factors['stimulus_type']:
                # For catch trials, distribute evenly across stimulus types
                stim_count = count // len(self.factors['stimulus_type'])
                if stim_count > 0:
                    stimulus_sequence.extend([stim_type] * stim_count)
            
            # Handle any remainders by adding random stimuli
            remaining = count - len(stimulus_sequence)
            if remaining > 0:
                stimulus_sequence.extend(np.random.choice(self.factors['stimulus_type'], size=remaining))
            
            # Shuffle the stimulus sequence
            np.random.shuffle(stimulus_sequence)
            
            # Add design details for each trial
            for idx, stimulus_type in enumerate(stimulus_sequence):
                # Compute trial number
                trial_idx = len(design)
                trial_number = 20 + trial_idx  # Offset of 20 trials
                
                # Choose a random jitter value (in ms)
                current_jitter = np.random.choice(self.factors['jitter'])
                
                # Get SOA condition index and actual value in milliseconds
                # For catch trials, SOA isn't meaningful but we assign a value for consistency
                soa_idx = idx % len(self.factors['soa_condition'])
                soa_value_ms = self.soa_values_ms[soa_idx]
                
                # Generate timestamps (either from CSV or calculated)
                if self.reference_csv_path is not None:
                    # Use CSV-based timestamps (linked to breathing audio)
                    base_ms = csv_milliseconds[csv_index % len(csv_milliseconds)]
                    csv_index += 1
                    jittered_ms = base_ms + current_jitter
                    soa_ms = jittered_ms + soa_value_ms
                    
                    # Format timestamps as "mm:ss.f"
                    timestamp_original = f"{int(base_ms // 60000):02}:{(base_ms % 60000) / 1000:.1f}"
                    timestamp_after_jitter = f"{int(jittered_ms // 60000):02}:{(jittered_ms % 60000) / 1000:.1f}"
                    timestamp_with_soa = f"{int(soa_ms // 60000):02}:{(soa_ms % 60000) / 1000:.1f}"
                    
                    # Store raw millisecond values for precise audio syncing
                    base_ms_raw = base_ms
                    jittered_ms_raw = jittered_ms
                    soa_ms_raw = soa_ms
                else:
                    # Use base_time and trial_duration method (regular intervals)
                    t_orig = self.base_time + pd.Timedelta(seconds=trial_number * self.trial_duration)
                    t_jitter = self.base_time + pd.Timedelta(seconds=(trial_number * self.trial_duration) + (current_jitter / 1000))
                    t_soa = self.base_time + pd.Timedelta(seconds=(trial_number * self.trial_duration) + (current_jitter / 1000) + (soa_value_ms / 1000))
                    
                    timestamp_original = t_orig.strftime('%M:%S.%f')[:-4]
                    timestamp_after_jitter = t_jitter.strftime('%M:%S.%f')[:-4]
                    timestamp_with_soa = t_soa.strftime('%M:%S.%f')[:-4]
                    
                    # Calculate raw millisecond values for precise audio syncing
                    base_ms_raw = trial_number * self.trial_duration * 1000
                    jittered_ms_raw = base_ms_raw + current_jitter
                    soa_ms_raw = jittered_ms_raw + soa_value_ms
                
                # For catch trials, there's no tactile component, so set is_tactile flag
                is_tactile = (trial_type != 'catch')
                
                # Add this trial to the design
                design.append({
                    'participant_id': participant_id,
                    'trial_number': trial_number,
                    'stimulus_type': stimulus_type,
                    'trial_type': trial_type,
                    'soa_condition_idx': soa_idx,
                    'soa_value_ms': soa_value_ms,  # Store exact SOA value in ms
                    'jitter_ms': current_jitter,
                    'timestamp_original': timestamp_original,
                    'timestamp_after_jitter': timestamp_after_jitter,
                    'timestamp_with_soa': timestamp_with_soa,  # New column with jitter + SOA
                    'base_ms': base_ms_raw,
                    'jittered_ms': jittered_ms_raw,
                    'soa_ms': soa_ms_raw,
                    'is_tactile': is_tactile  # Flag to indicate if trial has tactile component
                })
        
        # Convert to DataFrame
        design_df = pd.DataFrame(design)
        
        # Shuffle within each trial type to maintain balance
        shuffled_design = []
        for trial_type in self.trial_counts.keys():
            type_subset = design_df[design_df['trial_type'] == trial_type].copy()
            if not type_subset.empty:
                type_subset = type_subset.sample(frac=1, random_state=participant_id).reset_index(drop=True)
                shuffled_design.append(type_subset)
        
        # Recombine, sort by timestamp, and reset index
        design_df = pd.concat(shuffled_design).reset_index(drop=True)
        design_df = design_df.sort_values(by='timestamp_after_jitter').reset_index(drop=True)
        
        # Calculate and display expected experiment duration
        if EXPERIMENT_CONFIG['debug_mode'] and len(design_df) > 0:
            first_time = self._parse_timestamp(design_df['timestamp_original'].iloc[0])
            last_time = self._parse_timestamp(design_df['timestamp_original'].iloc[-1])
            duration_minutes = (last_time - first_time) / 60
            print(f"Participant {participant_id}: {len(design_df)} trials, duration ~{duration_minutes:.1f} minutes")
        
        return design_df
    
    def _parse_timestamp(self, timestamp_str):
        """Parse mm:ss.f timestamp to seconds"""
        parts = timestamp_str.split(':')
        if len(parts) == 2:
            minutes = int(parts[0])
            seconds = float(parts[1])
            return minutes * 60 + seconds
        return 0
    
    def save_designs(self, designs):
        """
        Save generated designs to CSV files.
        
        Args:
            designs: List of participant design DataFrames
        """
        for design in designs:
            if design.empty:
                continue
                
            participant_id = design['participant_id'].iloc[0]
            filename = f'participant_{participant_id}_design.csv'
            filepath = os.path.join(self.output_dir, filename)
            design.to_csv(filepath, index=False)
            
            if EXPERIMENT_CONFIG['debug_mode']:
                print(f"Saved design for participant {participant_id} with {len(design)} trials")
            else:
                print(f"Saved design for participant {participant_id}")
    
    def validate_designs(self, designs):
        """
        Comprehensively validate the generated designs to ensure balance.
        
        Args:
            designs: List of participant design DataFrames
        
        Returns:
            Dictionary with validation statistics
        """
        print("Starting design validation...")
        
        validation_results = {
            'stimulus_type': {},
            'trial_type': {},
            'soa_condition_idx': {},
            'jitter': {},
            'doegen_validation': {},
            'experiment_config': EXPERIMENT_CONFIG
        }
        
        try:
            # Combine all designs for overall validation
            all_designs = pd.concat(designs)
            
            # ------- Basic Distribution Validation -------
            # Validate each factor by grouping by trial type
            for factor in ['stimulus_type', 'trial_type', 'soa_condition_idx']:
                grouped_validation = all_designs.groupby(['trial_type', factor]).size() / len(designs)
                validation_results[factor] = grouped_validation.to_dict()
            
            print("Basic distribution validation complete.")
            
            # ------- Additional Dynamic Validation -------
            # Validate trial counts match configuration
            actual_counts = all_designs.groupby('trial_type').size().to_dict()
            validation_results['trial_counts'] = {
                'expected': self.trial_counts,
                'actual_per_participant': {k: v/len(designs) for k, v in actual_counts.items()}
            }
            
            # Validate SOA distribution
            soa_distribution = all_designs.groupby(['trial_type', 'soa_value_ms']).size().to_dict()
            validation_results['soa_distribution'] = soa_distribution
            
            # Find the approximate experiment duration
            if all_designs.shape[0] > 0:
                # Get last timestamp from each participant
                participant_durations = []
                for participant_id, group in all_designs.groupby('participant_id'):
                    if not group.empty:
                        last_time = group['timestamp_after_jitter'].iloc[-1]
                        parts = last_time.split(':')
                        if len(parts) == 2:
                            minutes = int(parts[0])
                            seconds = float(parts[1])
                            total_seconds = minutes * 60 + seconds
                            participant_durations.append(total_seconds)
                
                if participant_durations:
                    avg_duration = sum(participant_durations) / len(participant_durations)
                    validation_results['estimated_duration'] = {
                        'seconds': avg_duration,
                        'minutes': avg_duration / 60,
                        'per_trial': avg_duration / (sum(self.trial_counts.values()) / len(designs))
                    }
                    print(f"Estimated experiment duration: {avg_duration/60:.1f} minutes")
            
            # Save detailed validation report to JSON
            validation_report_path = os.path.join(self.output_dir, 'dynamic_validation_report.json')
            try:
                with open(validation_report_path, 'w') as f:
                    json.dump(validation_results, f, indent=4, default=str)
                print(f"Validation report saved to: {validation_report_path}")
            except Exception as e:
                print(f"Warning: Error saving validation report: {str(e)}")
            
            # Print summary of validation results
            print("\nValidation Results:")
            print(f"Total participants: {len(designs)}")
            print(f"Total trials per participant: {sum(self.trial_counts.values())}")
            print("Trial counts by condition:")
            for trial_type, count in self.trial_counts.items():
                actual = actual_counts.get(trial_type, 0) / len(designs)
                print(f"  {trial_type}: {count} expected, {actual:.1f} actual")
            
            if 'estimated_duration' in validation_results:
                print(f"Estimated duration: {validation_results['estimated_duration']['minutes']:.1f} minutes")
            
        except Exception as e:
            print(f"Error during validation: {str(e)}")
            import traceback
            traceback.print_exc()
            print("Validation could not be completed. Returning partial results.")
        
        print("Validation complete.")
        return validation_results

# ======================================================================
# AUDIO GENERATOR
# ======================================================================

class DynamicAudioGenerator:
    """
    Generates participant-specific audio files with embedded stimuli,
    automatically adjusting length based on trial counts.
    """
    
    def __init__(self, base_audio_path, output_dir, looming_stimuli_paths, tactile_stimulus_path, timestamps_csv_path=None):
        """
        Initialize the audio generator.
        
        Args:
            base_audio_path: Path to the base box breathing audio
            output_dir: Directory to save participant-specific audio files
            looming_stimuli_paths: Dictionary of paths to looming stimuli sounds
            tactile_stimulus_path: Path to the tactile stimulus sound
            timestamps_csv_path: Optional path to CSV with breathing timestamps
        """
        # Store stimulus paths
        self.looming_stimuli = looming_stimuli_paths
        self.tactile_stimulus_path = tactile_stimulus_path
        
        # Load base audio (breathing exercise)
        self.base_audio_data, self.sample_rate = sf.read(base_audio_path)
        
        # Store base audio info for dynamic extension
        self.base_audio_duration = len(self.base_audio_data) / self.sample_rate
        
        if EXPERIMENT_CONFIG['debug_mode']:
            print(f"Base audio duration: {self.base_audio_duration:.1f} seconds "
                  f"({self.base_audio_duration/60:.1f} minutes)")
        
        # Load tactile stimulus
        self.tactile_data, _ = sf.read(tactile_stimulus_path)
        
        # Load timestamps if provided
        self.timestamps_df = None
        if timestamps_csv_path:
            self.timestamps_df = pd.read_csv(timestamps_csv_path)
        
        # Output directory for audio files
        self.output_dir = output_dir
        os.makedirs(output_dir, exist_ok=True)
    
    def extend_base_audio(self, required_duration):
        """
        Dynamically extend the base audio to cover the required experiment duration.
        
        Args:
            required_duration: Required duration in seconds
            
        Returns:
            Extended audio data
        """
        # Calculate how many repetitions of the base audio we need
        repeats_needed = math.ceil(required_duration / self.base_audio_duration)
        
        if EXPERIMENT_CONFIG['debug_mode']:
            print(f"Extending base audio: needs {required_duration:.1f} seconds, "
                  f"using {repeats_needed} repetitions of base audio")
        
        # Tile the base audio to reach required duration
        if repeats_needed <= 1:
            # No extension needed
            return self.base_audio_data.copy()
        else:
            # Create an extended version
            extended_audio = np.tile(self.base_audio_data, (repeats_needed, 1))
            return extended_audio
    
    def generate_participant_audio(self, participant_id, design_df):
        """
        Generate two synchronized audio files for a participant.
        
        Args:
            participant_id: Participant identifier
            design_df: DataFrame with stimuli design
                
        Returns:
            Tuple (looming_audio, tactile_audio, injection_log)
        """
        # Check if design is empty
        if design_df.empty:
            print(f"Warning: Empty design for participant {participant_id}. Skipping audio generation.")
            return None, None, pd.DataFrame()
        
        # Find the required audio duration based on the latest timestamp
        latest_timestamp = design_df['timestamp_with_soa'].iloc[-1] if not design_df.empty else "00:00.0"
        latest_time_parts = latest_timestamp.split(':')
        minutes = int(latest_time_parts[0])
        seconds = float(latest_time_parts[1])
        required_duration = (minutes * 60 + seconds) + 30  # Add 30 second buffer
        
        if EXPERIMENT_CONFIG['debug_mode']:
            print(f"Required audio duration: {required_duration:.1f} seconds "
                  f"({required_duration/60:.1f} minutes)")
        
        # Extend base audio if needed
        extended_base_audio = self.extend_base_audio(required_duration)
        
        # Create two separate audio streams
        looming_audio = extended_base_audio.copy()  # Box breathing with looming sounds
        tactile_audio = np.zeros_like(looming_audio)  # Empty track matching looming shape
        
        # Ensure tactile stimulus is compatible format
        if self.tactile_data.ndim == 1 and looming_audio.ndim > 1:  # If tactile is mono and looming is stereo
            tactile_stereo = np.column_stack((self.tactile_data, self.tactile_data))
        elif self.tactile_data.ndim > 1 and looming_audio.ndim == 1:  # If tactile is stereo and looming is mono
            tactile_stereo = np.mean(self.tactile_data, axis=1)
        else:
            tactile_stereo = self.tactile_data
            
        # Initialize log to track all stimulus injections
        injection_log = []
        
        # Verify required columns
        required_columns = ['stimulus_type', 'jittered_ms', 'soa_ms', 'soa_value_ms']
        missing_columns = [col for col in required_columns if col not in design_df.columns]
        if missing_columns:
            print(f"Warning: Design DataFrame missing required columns: {missing_columns}. "
                  f"Cannot ensure precise synchronization.")
            return None, None, pd.DataFrame()
        
        # Process each trial in the design
        for _, row in design_df.iterrows():
            # Get exact millisecond timestamps from the design CSV
            jittered_ms = row['jittered_ms']  # For looming stimuli
            soa_ms = row['soa_ms']  # For tactile stimuli
            
            # Convert to sample indices
            looming_injection_point = int(jittered_ms / 1000 * self.sample_rate)
            tactile_injection_point = int(soa_ms / 1000 * self.sample_rate)
            
            # Get SOA value for logging
            soa_value_ms = row['soa_value_ms']
            
            # Load the appropriate looming stimulus based on design
            stim_type = row['stimulus_type']
            looming_path = self.looming_stimuli.get(stim_type)
            if looming_path is None:
                continue  # Skip if stimulus type not found
            
            # Read the looming stimulus audio file
            looming_data, _ = sf.read(looming_path)
            
            # Check if both stimuli fit within the audio bounds
            looming_fits = (looming_injection_point + len(looming_data) <= len(looming_audio))
            
            if looming_fits:
                # Inject looming stimulus into the first audio track
                try:
                    looming_audio[looming_injection_point:looming_injection_point+len(looming_data)] += looming_data
                except ValueError as e:
                    print(f"Error injecting looming stimulus: {e}, shapes: "
                          f"looming_audio={looming_audio.shape}, looming_data={looming_data.shape}")
                    continue
                
                # Inject tactile stimulus into the second audio track for non-catch trials
                if row.get('is_tactile', True) and row['trial_type'] != 'catch':
                    tactile_fits = (tactile_injection_point >= 0 and 
                                  tactile_injection_point + len(tactile_stereo) <= len(tactile_audio))
                    
                    if tactile_fits:
                        try:
                            tactile_audio[tactile_injection_point:tactile_injection_point+len(tactile_stereo)] += tactile_stereo
                        except ValueError as e:
                            print(f"Error injecting tactile stimulus: {e}, shapes: "
                                  f"tactile_audio={tactile_audio.shape}, tactile_stereo={tactile_stereo.shape}")
                            continue
                    
                # Log details of this injection
                injection_log.append({
                    'participant_id': participant_id,
                    'trial_number': row.get('trial_number', 0),
                    'stimulus_type': stim_type,
                    'soa_condition_idx': row.get('soa_condition_idx', 0),
                    'soa_value_ms': soa_value_ms,
                    'looming_injection_sample': looming_injection_point,
                    'tactile_injection_sample': tactile_injection_point,
                    'looming_time_ms': looming_injection_point / self.sample_rate * 1000,
                    'tactile_time_ms': tactile_injection_point / self.sample_rate * 1000,
                    'jittered_ms': jittered_ms,
                    'soa_ms': soa_ms,
                    'is_tactile': row.get('is_tactile', True)
                })
        
        return looming_audio, tactile_audio, pd.DataFrame(injection_log)
    
    def save_participant_files(self, participant_id, looming_audio, tactile_audio, injection_log, design_filename):
        """
        Save participant-specific audio and log files.
        
        Args:
            participant_id: Participant identifier
            looming_audio: Modified audio with looming stimuli
            tactile_audio: Audio track with tactile stimuli
            injection_log: DataFrame with injection details
            design_filename: Base design filename for naming consistency
        
        Returns:
            Tuple of saved file paths
        """
        # Check if audio data is None
        if looming_audio is None or tactile_audio is None:
            print(f"Warning: Cannot save audio for participant {participant_id} - missing audio data")
            return None, None, None
        
        # Use the design CSV base name (remove .csv extension)
        base_name = os.path.splitext(design_filename)[0]
        
        # Define filenames for output files
        looming_filename = f"{base_name}_looming.wav"
        tactile_filename = f"{base_name}_tactile.wav"
        log_filename = f"{base_name}_stimuli_log.csv"
        
        # Create full paths
        looming_path = os.path.join(self.output_dir, looming_filename)
        tactile_path = os.path.join(self.output_dir, tactile_filename)
        log_path = os.path.join(self.output_dir, log_filename)
        
        # Save the audio files
        sf.write(looming_path, looming_audio, self.sample_rate)
        sf.write(tactile_path, tactile_audio, self.sample_rate)
        
        # Save the detailed injection log (for analysis)
        if not injection_log.empty:
            injection_log.to_csv(log_path, index=False)
        
        print(f"Looming audio saved: {looming_filename}")
        print(f"Tactile audio saved: {tactile_filename}")
        print(f"Injection log saved: {log_filename}")
        
        return looming_path, tactile_path, log_path


# ======================================================================
# MAIN FUNCTION
# ======================================================================

def main():
    """
    Main function to run the dynamic PPS experiment setup process.
    """
    # Ensure output directories exist
    os.makedirs(DESIGN_OUTPUT_DIR, exist_ok=True)
    os.makedirs(AUDIO_OUTPUT_DIR, exist_ok=True)
    
    # Calculate and display trial counts
    trial_counts = calculate_trial_counts()
    total_trials = sum(trial_counts.values())
    
    # Display configuration overview
    print("\n===== PPS EXPERIMENT GENERATOR =====")
    print(f"Repetitions per condition: {EXPERIMENT_CONFIG['repetitions_per_condition']}")
    
    # Show which conditions are included
    active_conditions = [cond for cond, include in EXPERIMENT_CONFIG['included_conditions'].items() if include]
    print("Included conditions:", ", ".join(active_conditions))
    
    print(f"Catch trial percentage: {EXPERIMENT_CONFIG['catch_trial_percentage']}%")
    print(f"Total trials per participant: {total_trials}")
    print(f"Output directory: {DESIGN_OUTPUT_DIR}")
    
    # Step 1: Generate counterbalanced designs
    print("\n===== STEP 1: GENERATING COUNTERBALANCED DESIGNS =====")
    print(f"Using SOA conditions (ms): {SOA_CONDITIONS_MS}")
    
    counterbalancer = DynamicCounterbalancer(
        output_dir=DESIGN_OUTPUT_DIR,
        soa_conditions=SOA_CONDITIONS_MS,
        jitter_options=JITTER_OPTIONS_MS
    )
    
    designs = counterbalancer.generate_comprehensive_design(
        num_participants=NUM_PARTICIPANTS, 
        reference_csv_path=TIMESTAMPS_CSV_PATH
    )
    
    # Save designs to CSV files
    counterbalancer.save_designs(designs)
    
    # Run validation checks
    print("\n===== RUNNING VALIDATION CHECKS =====")
    validation_results = counterbalancer.validate_designs(designs)
    
    # Step 2: Generate participant-specific audio files
    print("\n===== STEP 2: GENERATING DYNAMIC AUDIO FILES =====")
    print(f"Creating TWO audio files per participant:")
    print(f"  - Looming audio: Box breathing with directional looming sounds")
    print(f"  - Tactile audio: Synchronized tactile stimuli with SOA offsets")
    
    # Initialize the audio generator
    audio_generator = DynamicAudioGenerator(
        base_audio_path=BASE_AUDIO_PATH,
        output_dir=AUDIO_OUTPUT_DIR,
        looming_stimuli_paths=LOOMING_STIMULI,
        tactile_stimulus_path=TACTILE_STIMULUS_PATH,
        timestamps_csv_path=TIMESTAMPS_CSV_PATH
    )
    
    # Process each design and generate audio pairs
    for i, design_df in enumerate(designs):
        if design_df.empty:
            continue
            
        participant_id = design_df['participant_id'].iloc[0]
        design_filename = f"participant_{participant_id}_design.csv"
        
        print(f"\nProcessing participant {participant_id} ({i+1}/{NUM_PARTICIPANTS})...")
        
        # Verify design CSV has necessary columns
        if 'jittered_ms' not in design_df.columns or 'soa_ms' not in design_df.columns:
            print(f"Warning: Design for participant {participant_id} missing critical timing columns.")
            print("Reading design from saved CSV file to ensure correct timing...")
            
            # Re-read the design CSV
            design_file_path = os.path.join(DESIGN_OUTPUT_DIR, design_filename)
            if os.path.exists(design_file_path):
                design_df = pd.read_csv(design_file_path)
                print("Successfully loaded design from CSV with all timing information.")
            else:
                print(f"Error: Cannot find design file {design_file_path}")
                continue
        
        # Generate audio for this participant
        try:
            looming_audio, tactile_audio, injection_log = audio_generator.generate_participant_audio(
                participant_id, 
                design_df
            )
            
            # Save both audio files and the log
            audio_generator.save_participant_files(
                participant_id, 
                looming_audio, 
                tactile_audio, 
                injection_log, 
                design_filename
            )
            
            # Verify synchronization
            print("Verifying synchronization...")
            if not injection_log.empty:
                for i, row in injection_log.head(3).iterrows():
                    looming_time = row['looming_time_ms']
                    tactile_time = row['tactile_time_ms']
                    expected_diff = row['soa_value_ms']
                    actual_diff = tactile_time - looming_time
                    
                    print(f"  Trial {i+1}: SOA expected={expected_diff}ms, actual={actual_diff:.2f}ms, " +
                          f"difference={abs(actual_diff-expected_diff):.4f}ms")
                
                if len(injection_log) > 3:
                    print(f"  ... and {len(injection_log) - 3} more trials")
            else:
                print("  Warning: No stimuli were injected - check design parameters")
                
        except Exception as e:
            print(f"Error generating audio for participant {participant_id}: {str(e)}")
            import traceback
            traceback.print_exc()
            continue
    
    # Print summary
    print(f"\n===== GENERATION COMPLETE =====")
    print(f"All designs generated and saved to: {DESIGN_OUTPUT_DIR}")
    print(f"All audio files generated and saved to: {AUDIO_OUTPUT_DIR}")
    print(f"Created {NUM_PARTICIPANTS} participant designs with:")
    for trial_type, count in trial_counts.items():
        print(f"  - {count} {trial_type} trials")
    
    if 'estimated_duration' in validation_results:
        print(f"Estimated experiment duration: {validation_results['estimated_duration']['minutes']:.1f} minutes per participant")
    
    print("Each participant has TWO audio files: one with looming stimuli and one with tactile stimuli")
    print("Process complete.")

if __name__ == "__main__":
    main()
```

```python
"""
Peripersonal Space (PPS) Experiment Generator

This script generates customizable PPS experiments with balanced conditions.
The experiment uses a pre-recorded box breathing audio track and inserts
stimuli at specific breathing phase markers indicated in a timestamp CSV.

The experiment design follows a factorial structure:
- Conditions: inhalation, exhalation, baseline (selectable)
- SOA values: 5 values [190, 400, 700, 1000, 1500] ms
- Stimulus directions: 3 directions (left, front, right)
- Repetitions: How many times each unique combination is repeated
- Catch trials: Added as a percentage of the total factorial design

Each repetition represents a full factorial cell of condition × SOA × direction.
"""

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import soundfile as sf
import scipy.signal as signal
from datetime import datetime
from itertools import product, combinations
import warnings
import json
import math
import random

# ======================================================================
# CONFIGURABLE PARAMETERS - MODIFY THESE TO CUSTOMIZE YOUR EXPERIMENT
# ======================================================================

# Experiment structure
EXPERIMENT_CONFIG = {
    'repetitions': 1,        # How many times to repeat the full factorial design
                             # (1 rep = 1 trial per condition×SOA×direction combination)
    
    'included_conditions': {
        'inhalation': True,  # Include inhalation trials
        'exhalation': False,  # Include exhalation trials
        'baseline': False,    # Include baseline trials
    },
    
    'catch_trial_percentage': 10,  # Additional catch trials as percentage of factorial trials
    'debug_mode': False,           # Enable for verbose output
}

# SOA conditions in milliseconds (Stimulus Onset Asynchrony)
SOA_CONDITIONS_MS = [190]

# Jitter options in milliseconds (temporal variability)
JITTER_OPTIONS_MS = [100, 200, 300, 400, 500]

# File paths - UPDATE THESE TO MATCH YOUR SYSTEM
BASE_DIR = r"C:\Users\cogpsy-vrlab\Documents\GitHub\BreathingSpace"

# Original audio paths (update if needed)
ORIG_AUDIO_DIR = r"C:\Users\cogpsy-vrlab\Documents\PPS_module\BreathingPilot"

# Box breathing audio
BASE_AUDIO_PATH = os.path.join(ORIG_AUDIO_DIR, "AudioInstructionsBoxBreathing", "TargetStimuli", 
                              "6. Box Breathing Combined (Segment Injected).wav")

# Timestamp reference for breathing phases
TIMESTAMPS_CSV_PATH = os.path.join(ORIG_AUDIO_DIR, "AudioInstructionsBoxBreathing", "TargetStimuli", 
                                  "BoxBreathing_extended_timestamps.csv")

# Output directories
DESIGN_OUTPUT_DIR = os.path.join(BASE_DIR, "ExperimentLog")
AUDIO_OUTPUT_DIR = os.path.join(BASE_DIR, "ExperimentAudio")

# Looming stimuli (directional audio)
LOOMING_STIMULI = {
    'right': os.path.join(ORIG_AUDIO_DIR, "PPS_Experiment_Module", "right_az90_FABIAN_HRIR_natural.wav"),
    'front': os.path.join(ORIG_AUDIO_DIR, "PPS_Experiment_Module", "front_az0_FABIAN_HRIR_natural.wav"),
    'left': os.path.join(ORIG_AUDIO_DIR, "PPS_Experiment_Module", "left_az-90_FABIAN_HRIR_natural.wav")
}

# Tactile stimulus (to be synchronized with looming stimuli + SOA)
TACTILE_STIMULUS_PATH = os.path.join(ORIG_AUDIO_DIR, "PPS_Experiment_Module", "tactile_stimulus.wav")

# Number of participants to generate designs for
NUM_PARTICIPANTS = 100

# ======================================================================
# DYNAMIC TRIAL COUNT CALCULATOR
# ======================================================================

def calculate_trial_counts():
    """
    Calculate trial counts based on the experiment configuration.
    
    Returns:
        Dictionary with trial counts for each condition
    """
    # Get configuration
    reps = EXPERIMENT_CONFIG['repetitions']
    stimuli_directions = 3  # left, front, right
    soa_values = len(SOA_CONDITIONS_MS)  # 5 SOA values
    
    # Get selected conditions
    selected_conditions = [cond for cond, include in EXPERIMENT_CONFIG['included_conditions'].items() if include]
    
    # Each condition has trials for each combination of direction, SOA, and repetition
    # trials_per_condition = reps * stimuli_directions * soa_values
    
    # For clearer organization, we'll calculate for each condition separately
    basic_trial_counts = {}
    for cond in selected_conditions:
        basic_trial_counts[cond] = reps * stimuli_directions * soa_values
    
    # Calculate total number of factorial design trials
    total_factorial_trials = sum(basic_trial_counts.values())
    
    # Calculate catch trials based on percentage
    catch_percentage = EXPERIMENT_CONFIG['catch_trial_percentage']
    catch_trials = round(total_factorial_trials * (catch_percentage / 100))
    
    # Add catch trials to the count
    final_counts = basic_trial_counts.copy()
    final_counts['catch'] = catch_trials
    
    if EXPERIMENT_CONFIG['debug_mode']:
        print(f"Trial count calculation:")
        print(f"  Repetitions: {reps}")
        print(f"  SOA values: {soa_values}")
        print(f"  Stimulus directions: {stimuli_directions}")
        print(f"  Conditions: {', '.join(selected_conditions)}")
        print(f"  Factorial combinations: {reps} × {soa_values} × {stimuli_directions} × {len(selected_conditions)} = {total_factorial_trials}")
        print(f"  Catch trials ({catch_percentage}%): {catch_trials}")
        print(f"  Total trials: {sum(final_counts.values())}")
        
        print(f"\nTrial counts by condition:")
        for cond, count in final_counts.items():
            print(f"  {cond}: {count} trials")
    
    return final_counts

# ======================================================================
# COUNTERBALANCED DESIGN GENERATOR
# ======================================================================

class DynamicCounterbalancer:
    """
    Creates counterbalanced experimental designs with flexible trial counts
    based on configuration parameters.
    """
    
    def __init__(self, output_dir, soa_conditions, jitter_options):
        """
        Initialize the counterbalancer with experiment parameters.
        
        Args:
            output_dir: Directory to save design files
            soa_conditions: List of SOA values in milliseconds
            jitter_options: List of jitter values in milliseconds
        """
        # Ensure output directory exists
        self.output_dir = output_dir
        os.makedirs(output_dir, exist_ok=True)
        
        # Calculate trial counts based on configuration
        self.trial_counts = calculate_trial_counts()
        
        # Define experimental factors
        self.factors = {
            'stimulus_type': ['right', 'front', 'left'],  # Stimulus directions
            'trial_type': list(self.trial_counts.keys()),  # Active trial types
            'soa_condition': list(range(len(soa_conditions))),  # SOA condition indices
            'jitter': jitter_options  # Jitter in milliseconds
        }
        
        # Store actual SOA values for reference
        self.soa_values_ms = soa_conditions
        
        # Default timing parameters
        self.base_time = pd.Timestamp('2025-03-01 09:00:00')
        self.trial_duration = 8  # Average seconds between trials
        
        # This will be set if a reference CSV is provided
        self.reference_csv_path = None

    def generate_comprehensive_design(self, num_participants=100, reference_csv_path=None):
        """
        Generate a counterbalanced design for multiple participants.
        
        Args:
            num_participants: Number of participant designs to generate
            reference_csv_path: Path to CSV with breathing exercise timestamps
        
        Returns:
            List of DataFrames, one for each participant
        """
        self.reference_csv_path = reference_csv_path
        designs = []
        
        # Set a fixed seed for reproducibility
        np.random.seed(42)
        
        # Generate design for each participant
        for participant_id in range(num_participants):
            design = self._generate_single_participant_design(participant_id)
            designs.append(design)
        
        return designs
    
    def _generate_factorial_combinations(self, trial_type):
        """
        Generate all factorial combinations for a given trial type.
        
        For each trial type, creates all combinations of:
        - SOA values
        - Stimulus directions
        - Repetitions
        
        Args:
            trial_type: The trial type (inhalation, exhalation, baseline, catch)
            
        Returns:
            List of dictionaries, each representing a trial
        """
        combinations = []
        
        # For catch trials, we handle them differently
        if trial_type == 'catch':
            catch_count = self.trial_counts['catch']
            
            # For catch trials, create balanced direction distribution
            for i in range(catch_count):
                # Create a catch trial with random direction and SOA
                # (SOA doesn't matter for catch trials, but we assign one for consistency)
                direction = self.factors['stimulus_type'][i % len(self.factors['stimulus_type'])]
                soa_idx = i % len(self.factors['soa_condition'])
                
                combinations.append({
                    'trial_type': trial_type,
                    'stimulus_type': direction,
                    'soa_condition_idx': soa_idx,
                    'soa_value_ms': self.soa_values_ms[soa_idx],
                    'is_tactile': False  # Catch trials don't have tactile component
                })
            
        else:
            # For regular trials, create full factorial design
            reps = EXPERIMENT_CONFIG['repetitions']
            
            # For each repetition
            for rep in range(reps):
                # For each SOA value
                for soa_idx, soa_value in enumerate(self.soa_values_ms):
                    # For each stimulus direction
                    for direction in self.factors['stimulus_type']:
                        combinations.append({
                            'trial_type': trial_type,
                            'stimulus_type': direction,
                            'soa_condition_idx': soa_idx,
                            'soa_value_ms': soa_value,
                            'is_tactile': True  # Regular trials have tactile component
                        })
        
        # Shuffle the combinations to avoid sequential patterns
        random.shuffle(combinations)
        return combinations
    
    def _generate_single_participant_design(self, participant_id):
        """
        Generate a balanced design for a single participant with precise timestamps.
        
        Args:
            participant_id: Unique identifier for the participant
        
        Returns:
            DataFrame with complete experimental design including timestamps
        """
        design = []
        
        # If a reference CSV is provided, use it for timing
        if self.reference_csv_path is not None:
            timestamp_df = pd.read_csv(self.reference_csv_path)
            csv_milliseconds = timestamp_df['milliseconds'].astype(int).values
            total_trials = sum(self.trial_counts.values())
            
            # Handle case where there aren't enough timestamps
            if total_trials > len(csv_milliseconds):
                if EXPERIMENT_CONFIG['debug_mode']:
                    print(f"Warning: Not enough timestamps in CSV. Needed {total_trials}, found {len(csv_milliseconds)}.")
                    print(f"Will recycle timestamps as needed.")
                
                # Create a repeated version of timestamps to cover all trials
                repeats_needed = math.ceil(total_trials / len(csv_milliseconds))
                csv_milliseconds = np.tile(csv_milliseconds, repeats_needed)[:total_trials]
            
            csv_index = 0  # Track which CSV row to use for each trial
        
        # Generate trials for each active condition
        all_trials = []
        for trial_type, count in self.trial_counts.items():
            if count > 0:
                # Generate factorial combinations for this trial type
                type_trials = self._generate_factorial_combinations(trial_type)
                all_trials.extend(type_trials)
        
        # Shuffle all trials to ensure mixed presentation
        random.shuffle(all_trials)
        
        # Add timing information to each trial
        for trial_idx, trial in enumerate(all_trials):
            # Compute trial number (starting after an initial offset)
            trial_number = 20 + trial_idx  # Offset of 20 trials
            
            # Choose a random jitter value (in ms)
            current_jitter = np.random.choice(self.factors['jitter'])
            
            # Get trial information
            trial_type = trial['trial_type']
            stimulus_type = trial['stimulus_type']
            soa_idx = trial['soa_condition_idx']
            soa_value_ms = trial['soa_value_ms']
            is_tactile = trial['is_tactile']
            
            # Generate timestamps (either from CSV or calculated)
            if self.reference_csv_path is not None:
                # Use CSV-based timestamps (linked to breathing audio)
                base_ms = csv_milliseconds[csv_index % len(csv_milliseconds)]
                csv_index += 1
                jittered_ms = base_ms + current_jitter
                soa_ms = jittered_ms + soa_value_ms
                
                # Format timestamps as "mm:ss.f"
                timestamp_original = f"{int(base_ms // 60000):02}:{(base_ms % 60000) / 1000:.1f}"
                timestamp_after_jitter = f"{int(jittered_ms // 60000):02}:{(jittered_ms % 60000) / 1000:.1f}"
                timestamp_with_soa = f"{int(soa_ms // 60000):02}:{(soa_ms % 60000) / 1000:.1f}"
                
                # Store raw millisecond values for precise audio syncing
                base_ms_raw = base_ms
                jittered_ms_raw = jittered_ms
                soa_ms_raw = soa_ms
            else:
                # Use base_time and trial_duration method (regular intervals)
                t_orig = self.base_time + pd.Timedelta(seconds=trial_number * self.trial_duration)
                t_jitter = self.base_time + pd.Timedelta(seconds=(trial_number * self.trial_duration) + (current_jitter / 1000))
                t_soa = self.base_time + pd.Timedelta(seconds=(trial_number * self.trial_duration) + (current_jitter / 1000) + (soa_value_ms / 1000))
                
                timestamp_original = t_orig.strftime('%M:%S.%f')[:-4]
                timestamp_after_jitter = t_jitter.strftime('%M:%S.%f')[:-4]
                timestamp_with_soa = t_soa.strftime('%M:%S.%f')[:-4]
                
                # Calculate raw millisecond values for precise audio syncing
                base_ms_raw = trial_number * self.trial_duration * 1000
                jittered_ms_raw = base_ms_raw + current_jitter
                soa_ms_raw = jittered_ms_raw + soa_value_ms
            
            # Add this trial to the design
            design.append({
                'participant_id': participant_id,
                'trial_number': trial_number,
                'stimulus_type': stimulus_type,
                'trial_type': trial_type,
                'soa_condition_idx': soa_idx,
                'soa_value_ms': soa_value_ms,  # Store exact SOA value in ms
                'jitter_ms': current_jitter,
                'timestamp_original': timestamp_original,
                'timestamp_after_jitter': timestamp_after_jitter,
                'timestamp_with_soa': timestamp_with_soa,  # New column with jitter + SOA
                'base_ms': base_ms_raw,
                'jittered_ms': jittered_ms_raw,
                'soa_ms': soa_ms_raw,
                'is_tactile': is_tactile  # Flag to indicate if trial has tactile component
            })
        
        # Convert to DataFrame and sort by timestamp
        design_df = pd.DataFrame(design)
        if not design_df.empty:
            design_df = design_df.sort_values(by='timestamp_after_jitter').reset_index(drop=True)
        
            # Calculate and display expected experiment duration
            if EXPERIMENT_CONFIG['debug_mode']:
                first_time = self._parse_timestamp(design_df['timestamp_original'].iloc[0])
                last_time = self._parse_timestamp(design_df['timestamp_original'].iloc[-1])
                duration_minutes = (last_time - first_time) / 60
                print(f"Participant {participant_id}: {len(design_df)} trials, duration ~{duration_minutes:.1f} minutes")
        
        return design_df
    
    def _parse_timestamp(self, timestamp_str):
        """Parse mm:ss.f timestamp to seconds"""
        parts = timestamp_str.split(':')
        if len(parts) == 2:
            minutes = int(parts[0])
            seconds = float(parts[1])
            return minutes * 60 + seconds
        return 0
    
    def save_designs(self, designs):
        """
        Save generated designs to CSV files.
        
        Args:
            designs: List of participant design DataFrames
        """
        for design in designs:
            if design.empty:
                continue
                
            participant_id = design['participant_id'].iloc[0]
            filename = f'participant_{participant_id}_design.csv'
            filepath = os.path.join(self.output_dir, filename)
            design.to_csv(filepath, index=False)
            
            if EXPERIMENT_CONFIG['debug_mode']:
                print(f"Saved design for participant {participant_id} with {len(design)} trials")
            else:
                print(f"Saved design for participant {participant_id}")
    
    def validate_designs(self, designs):
        """
        Comprehensively validate the generated designs to ensure balance.
        
        Args:
            designs: List of participant design DataFrames
        
        Returns:
            Dictionary with validation statistics
        """
        print("Starting design validation...")
        
        validation_results = {
            'stimulus_type': {},
            'trial_type': {},
            'soa_condition_idx': {},
            'jitter': {},
            'doegen_validation': {},
            'experiment_config': EXPERIMENT_CONFIG
        }
        
        try:
            # Combine all designs for overall validation
            all_designs = pd.concat(designs)
            
            # ------- Basic Distribution Validation -------
            # Validate each factor by grouping by trial type
            for factor in ['stimulus_type', 'trial_type', 'soa_condition_idx']:
                grouped_validation = all_designs.groupby(['trial_type', factor]).size() / len(designs)
                validation_results[factor] = grouped_validation.to_dict()
            
            print("Basic distribution validation complete.")
            
            # ------- Additional Factorial Validation -------
            # For non-catch trials, validate the factorial structure
            non_catch = all_designs[all_designs['trial_type'] != 'catch']
            
            # Count occurrences of each trial_type × soa × stimulus_type combination
            factorial_counts = non_catch.groupby(['trial_type', 'soa_condition_idx', 'stimulus_type']).size()
            
            # Should have equal counts for all combinations
            factorial_stats = {
                'mean': factorial_counts.mean(),
                'std': factorial_counts.std(),
                'min': factorial_counts.min(),
                'max': factorial_counts.max(),
                'expected': EXPERIMENT_CONFIG['repetitions']
            }
            
            validation_results['factorial_design'] = {
                'statistics': factorial_stats,
                'counts': factorial_counts.to_dict()
            }
            
            if EXPERIMENT_CONFIG['debug_mode']:
                print("Factorial design validation:")
                print(f"  Expected repetitions: {EXPERIMENT_CONFIG['repetitions']}")
                print(f"  Actual mean: {factorial_stats['mean']:.2f}")
                print(f"  Std dev: {factorial_stats['std']:.2f}")
                print(f"  Range: {factorial_stats['min']} - {factorial_stats['max']}")
            
            # ------- Additional Dynamic Validation -------
            # Validate trial counts match configuration
            actual_counts = all_designs.groupby('trial_type').size().to_dict()
            validation_results['trial_counts'] = {
                'expected': self.trial_counts,
                'actual_per_participant': {k: v/len(designs) for k, v in actual_counts.items()}
            }
            
            # Validate SOA distribution
            soa_distribution = all_designs.groupby(['trial_type', 'soa_value_ms']).size().to_dict()
            validation_results['soa_distribution'] = soa_distribution
            
            # Find the approximate experiment duration
            if all_designs.shape[0] > 0:
                # Get last timestamp from each participant
                participant_durations = []
                for participant_id, group in all_designs.groupby('participant_id'):
                    if not group.empty:
                        last_time = group['timestamp_after_jitter'].iloc[-1]
                        parts = last_time.split(':')
                        if len(parts) == 2:
                            minutes = int(parts[0])
                            seconds = float(parts[1])
                            total_seconds = minutes * 60 + seconds
                            participant_durations.append(total_seconds)
                
                if participant_durations:
                    avg_duration = sum(participant_durations) / len(participant_durations)
                    validation_results['estimated_duration'] = {
                        'seconds': avg_duration,
                        'minutes': avg_duration / 60,
                        'per_trial': avg_duration / (sum(self.trial_counts.values()) / len(designs))
                    }
                    print(f"Estimated experiment duration: {avg_duration/60:.1f} minutes")
            
            # Save detailed validation report to JSON
            validation_report_path = os.path.join(self.output_dir, 'dynamic_validation_report.json')
            try:
                with open(validation_report_path, 'w') as f:
                    json.dump(validation_results, f, indent=4, default=str)
                print(f"Validation report saved to: {validation_report_path}")
            except Exception as e:
                print(f"Warning: Error saving validation report: {str(e)}")
            
            # Print summary of validation results
            print("\nValidation Results:")
            print(f"Total participants: {len(designs)}")
            print(f"Total trials per participant: {sum(self.trial_counts.values())}")
            print("Trial counts by condition:")
            for trial_type, count in self.trial_counts.items():
                actual = actual_counts.get(trial_type, 0) / len(designs)
                print(f"  {trial_type}: {count} expected, {actual:.1f} actual")
            
            if 'estimated_duration' in validation_results:
                print(f"Estimated duration: {validation_results['estimated_duration']['minutes']:.1f} minutes")
            
        except Exception as e:
            print(f"Error during validation: {str(e)}")
            import traceback
            traceback.print_exc()
            print("Validation could not be completed. Returning partial results.")
        
        print("Validation complete.")
        return validation_results

# ======================================================================
# AUDIO GENERATOR
# ======================================================================

class DynamicAudioGenerator:
    """
    Generates participant-specific audio files with embedded stimuli,
    automatically adjusting length based on trial counts.
    """
    
    def __init__(self, base_audio_path, output_dir, looming_stimuli_paths, tactile_stimulus_path, timestamps_csv_path=None):
        """
        Initialize the audio generator.
        
        Args:
            base_audio_path: Path to the base box breathing audio
            output_dir: Directory to save participant-specific audio files
            looming_stimuli_paths: Dictionary of paths to looming stimuli sounds
            tactile_stimulus_path: Path to the tactile stimulus sound
            timestamps_csv_path: Optional path to CSV with breathing timestamps
        """
        # Store stimulus paths
        self.looming_stimuli = looming_stimuli_paths
        self.tactile_stimulus_path = tactile_stimulus_path
        
        # Load base audio (breathing exercise)
        self.base_audio_data, self.sample_rate = sf.read(base_audio_path)
        
        # Store base audio info for dynamic extension
        self.base_audio_duration = len(self.base_audio_data) / self.sample_rate
        
        if EXPERIMENT_CONFIG['debug_mode']:
            print(f"Base audio duration: {self.base_audio_duration:.1f} seconds "
                  f"({self.base_audio_duration/60:.1f} minutes)")
        
        # Load tactile stimulus
        self.tactile_data, _ = sf.read(tactile_stimulus_path)
        
        # Load timestamps if provided
        self.timestamps_df = None
        if timestamps_csv_path:
            self.timestamps_df = pd.read_csv(timestamps_csv_path)
        
        # Output directory for audio files
        self.output_dir = output_dir
        os.makedirs(output_dir, exist_ok=True)
    
    def extend_base_audio(self, required_duration):
        """
        Dynamically extend the base audio to cover the required experiment duration.
        
        Args:
            required_duration: Required duration in seconds
            
        Returns:
            Extended audio data
        """
        # Calculate how many repetitions of the base audio we need
        repeats_needed = math.ceil(required_duration / self.base_audio_duration)
        
        if EXPERIMENT_CONFIG['debug_mode']:
            print(f"Extending base audio: needs {required_duration:.1f} seconds, "
                  f"using {repeats_needed} repetitions of base audio")
        
        # Tile the base audio to reach required duration
        if repeats_needed <= 1:
            # No extension needed
            return self.base_audio_data.copy()
        else:
            # Create an extended version
            extended_audio = np.tile(self.base_audio_data, (repeats_needed, 1))
            return extended_audio
    
    def generate_participant_audio(self, participant_id, design_df):
        """
        Generate two synchronized audio files for a participant.
        
        Args:
            participant_id: Participant identifier
            design_df: DataFrame with stimuli design
                
        Returns:
            Tuple (looming_audio, tactile_audio, injection_log)
        """
        # Check if design is empty
        if design_df.empty:
            print(f"Warning: Empty design for participant {participant_id}. Skipping audio generation.")
            return None, None, pd.DataFrame()
        
        # Find the required audio duration based on the latest timestamp
        latest_timestamp = design_df['timestamp_with_soa'].iloc[-1] if not design_df.empty else "00:00.0"
        latest_time_parts = latest_timestamp.split(':')
        minutes = int(latest_time_parts[0])
        seconds = float(latest_time_parts[1])
        required_duration = (minutes * 60 + seconds) + 30  # Add 30 second buffer
        
        if EXPERIMENT_CONFIG['debug_mode']:
            print(f"Required audio duration: {required_duration:.1f} seconds "
                  f"({required_duration/60:.1f} minutes)")
        
        # Extend base audio if needed
        extended_base_audio = self.extend_base_audio(required_duration)
        
        # Create two separate audio streams
        looming_audio = extended_base_audio.copy()  # Box breathing with looming sounds
        tactile_audio = np.zeros_like(looming_audio)  # Empty track matching looming shape
        
        # Ensure tactile stimulus is compatible format
        if self.tactile_data.ndim == 1 and looming_audio.ndim > 1:  # If tactile is mono and looming is stereo
            tactile_stereo = np.column_stack((self.tactile_data, self.tactile_data))
        elif self.tactile_data.ndim > 1 and looming_audio.ndim == 1:  # If tactile is stereo and looming is mono
            tactile_stereo = np.mean(self.tactile_data, axis=1)
        else:
            tactile_stereo = self.tactile_data
            
        # Initialize log to track all stimulus injections
        injection_log = []
        
        # Verify required columns
        required_columns = ['stimulus_type', 'jittered_ms', 'soa_ms', 'soa_value_ms']
        missing_columns = [col for col in required_columns if col not in design_df.columns]
        if missing_columns:
            print(f"Warning: Design DataFrame missing required columns: {missing_columns}. "
                  f"Cannot ensure precise synchronization.")
            return None, None, pd.DataFrame()
        
        # Process each trial in the design
        for _, row in design_df.iterrows():
            # Get exact millisecond timestamps from the design CSV
            jittered_ms = row['jittered_ms']  # For looming stimuli
            soa_ms = row['soa_ms']  # For tactile stimuli
            
            # Convert to sample indices
            looming_injection_point = int(jittered_ms / 1000 * self.sample_rate)
            tactile_injection_point = int(soa_ms / 1000 * self.sample_rate)
            
            # Get SOA value for logging
            soa_value_ms = row['soa_value_ms']
            
            # Load the appropriate looming stimulus based on design
            stim_type = row['stimulus_type']
            looming_path = self.looming_stimuli.get(stim_type)
            if looming_path is None:
                continue  # Skip if stimulus type not found
            
            # Read the looming stimulus audio file
            looming_data, _ = sf.read(looming_path)
            
            # Check if both stimuli fit within the audio bounds
            looming_fits = (looming_injection_point + len(looming_data) <= len(looming_audio))
            
            if looming_fits:
                # Inject looming stimulus into the first audio track
                try:
                    looming_audio[looming_injection_point:looming_injection_point+len(looming_data)] += looming_data
                except ValueError as e:
                    print(f"Error injecting looming stimulus: {e}, shapes: "
                          f"looming_audio={looming_audio.shape}, looming_data={looming_data.shape}")
                    continue
                
                # Inject tactile stimulus into the second audio track for non-catch trials
                if row.get('is_tactile', True) and row['trial_type'] != 'catch':
                    tactile_fits = (tactile_injection_point >= 0 and 
                                  tactile_injection_point + len(tactile_stereo) <= len(tactile_audio))
                    
                    if tactile_fits:
                        try:
                            tactile_audio[tactile_injection_point:tactile_injection_point+len(tactile_stereo)] += tactile_stereo
                        except ValueError as e:
                            print(f"Error injecting tactile stimulus: {e}, shapes: "
                                  f"tactile_audio={tactile_audio.shape}, tactile_stereo={tactile_stereo.shape}")
                            continue
                    
                # Log details of this injection
                injection_log.append({
                    'participant_id': participant_id,
                    'trial_number': row.get('trial_number', 0),
                    'stimulus_type': stim_type,
                    'soa_condition_idx': row.get('soa_condition_idx', 0),
                    'soa_value_ms': soa_value_ms,
                    'looming_injection_sample': looming_injection_point,
                    'tactile_injection_sample': tactile_injection_point,
                    'looming_time_ms': looming_injection_point / self.sample_rate * 1000,
                    'tactile_time_ms': tactile_injection_point / self.sample_rate * 1000,
                    'jittered_ms': jittered_ms,
                    'soa_ms': soa_ms,
                    'is_tactile': row.get('is_tactile', True)
                })
        
        return looming_audio, tactile_audio, pd.DataFrame(injection_log)
    
    def save_participant_files(self, participant_id, looming_audio, tactile_audio, injection_log, design_filename):
        """
        Save participant-specific audio and log files.
        
        Args:
            participant_id: Participant identifier
            looming_audio: Modified audio with looming stimuli
            tactile_audio: Audio track with tactile stimuli
            injection_log: DataFrame with injection details
            design_filename: Base design filename for naming consistency
        
        Returns:
            Tuple of saved file paths
        """
        # Check if audio data is None
        if looming_audio is None or tactile_audio is None:
            print(f"Warning: Cannot save audio for participant {participant_id} - missing audio data")
            return None, None, None
        
        # Use the design CSV base name (remove .csv extension)
        base_name = os.path.splitext(design_filename)[0]
        
        # Define filenames for output files
        looming_filename = f"{base_name}_looming.wav"
        tactile_filename = f"{base_name}_tactile.wav"
        log_filename = f"{base_name}_stimuli_log.csv"
        
        # Create full paths
        looming_path = os.path.join(self.output_dir, looming_filename)
        tactile_path = os.path.join(self.output_dir, tactile_filename)
        log_path = os.path.join(self.output_dir, log_filename)
        
        # Save the audio files
        sf.write(looming_path, looming_audio, self.sample_rate)
        sf.write(tactile_path, tactile_audio, self.sample_rate)
        
        # Save the detailed injection log (for analysis)
        if not injection_log.empty:
            injection_log.to_csv(log_path, index=False)
        
        print(f"Looming audio saved: {looming_filename}")
        print(f"Tactile audio saved: {tactile_filename}")
        print(f"Injection log saved: {log_filename}")
        
        return looming_path, tactile_path, log_path


# ======================================================================
# MAIN FUNCTION
# ======================================================================

def main():
    """
    Main function to run the PPS experiment setup process.
    """
    # Ensure output directories exist
    os.makedirs(DESIGN_OUTPUT_DIR, exist_ok=True)
    os.makedirs(AUDIO_OUTPUT_DIR, exist_ok=True)
    
    # Calculate factorial structure and trial counts
    trial_counts = calculate_trial_counts()
    total_trials = sum(trial_counts.values())
    
    # Display configuration overview
    print("\n===== PPS EXPERIMENT GENERATOR =====")
    print(f"Factorial design with {EXPERIMENT_CONFIG['repetitions']} repetitions per combination")
    
    # Calculate factorial combinations
    active_conditions = [cond for cond, include in EXPERIMENT_CONFIG['included_conditions'].items() if include]
    soa_values = len(SOA_CONDITIONS_MS)
    stim_directions = 3
    factorial_trials = EXPERIMENT_CONFIG['repetitions'] * len(active_conditions) * soa_values * stim_directions
    
    print(f"Each repetition includes:")
    print(f"  • {len(active_conditions)} conditions ({', '.join(active_conditions)})")
    print(f"  • {soa_values} SOA values {SOA_CONDITIONS_MS}")
    print(f"  • {stim_directions} stimulus directions (left, front, right)")
    print(f"  = {factorial_trials} total factorial trials")
    print(f"  + {trial_counts['catch']} catch trials ({EXPERIMENT_CONFIG['catch_trial_percentage']}%)")
    print(f"  = {total_trials} total trials per participant")
    
    print(f"Output directory: {DESIGN_OUTPUT_DIR}")
    
    # Step 1: Generate counterbalanced designs
    print("\n===== STEP 1: GENERATING COUNTERBALANCED DESIGNS =====")
    print(f"Using SOA conditions (ms): {SOA_CONDITIONS_MS}")
    
    counterbalancer = DynamicCounterbalancer(
        output_dir=DESIGN_OUTPUT_DIR,
        soa_conditions=SOA_CONDITIONS_MS,
        jitter_options=JITTER_OPTIONS_MS
    )
    
    designs = counterbalancer.generate_comprehensive_design(
        num_participants=NUM_PARTICIPANTS, 
        reference_csv_path=TIMESTAMPS_CSV_PATH
    )
    
    # Save designs to CSV files
    counterbalancer.save_designs(designs)
    
    # Run validation checks
    print("\n===== RUNNING VALIDATION CHECKS =====")
    validation_results = counterbalancer.validate_designs(designs)
    
    # Step 2: Generate participant-specific audio files
    print("\n===== STEP 2: GENERATING DYNAMIC AUDIO FILES =====")
    print(f"Creating TWO audio files per participant:")
    print(f"  - Looming audio: Box breathing with directional looming sounds")
    print(f"  - Tactile audio: Synchronized tactile stimuli with SOA offsets")
    
    # Initialize the audio generator
    audio_generator = DynamicAudioGenerator(
        base_audio_path=BASE_AUDIO_PATH,
        output_dir=AUDIO_OUTPUT_DIR,
        looming_stimuli_paths=LOOMING_STIMULI,
        tactile_stimulus_path=TACTILE_STIMULUS_PATH,
        timestamps_csv_path=TIMESTAMPS_CSV_PATH
    )
    
    # Process each design and generate audio pairs
    for i, design_df in enumerate(designs):
        if design_df.empty:
            continue
            
        participant_id = design_df['participant_id'].iloc[0]
        design_filename = f"participant_{participant_id}_design.csv"
        
        print(f"\nProcessing participant {participant_id} ({i+1}/{NUM_PARTICIPANTS})...")
        
        # Verify design CSV has necessary columns
        if 'jittered_ms' not in design_df.columns or 'soa_ms' not in design_df.columns:
            print(f"Warning: Design for participant {participant_id} missing critical timing columns.")
            print("Reading design from saved CSV file to ensure correct timing...")
            
            # Re-read the design CSV
            design_file_path = os.path.join(DESIGN_OUTPUT_DIR, design_filename)
            if os.path.exists(design_file_path):
                design_df = pd.read_csv(design_file_path)
                print("Successfully loaded design from CSV with all timing information.")
            else:
                print(f"Error: Cannot find design file {design_file_path}")
                continue
        
        # Generate audio for this participant
        try:
            looming_audio, tactile_audio, injection_log = audio_generator.generate_participant_audio(
                participant_id, 
                design_df
            )
            
            # Save both audio files and the log
            audio_generator.save_participant_files(
                participant_id, 
                looming_audio, 
                tactile_audio, 
                injection_log, 
                design_filename
            )
            
            # Verify synchronization
            print("Verifying synchronization...")
            if not injection_log.empty:
                for i, row in injection_log.head(3).iterrows():
                    looming_time = row['looming_time_ms']
                    tactile_time = row['tactile_time_ms']
                    expected_diff = row['soa_value_ms']
                    actual_diff = tactile_time - looming_time
                    
                    print(f"  Trial {i+1}: SOA expected={expected_diff}ms, actual={actual_diff:.2f}ms, " +
                          f"difference={abs(actual_diff-expected_diff):.4f}ms")
                
                if len(injection_log) > 3:
                    print(f"  ... and {len(injection_log) - 3} more trials")
            else:
                print("  Warning: No stimuli were injected - check design parameters")
                
        except Exception as e:
            print(f"Error generating audio for participant {participant_id}: {str(e)}")
            import traceback
            traceback.print_exc()
            continue
    
    # Print summary
    print(f"\n===== GENERATION COMPLETE =====")
    print(f"All designs generated and saved to: {DESIGN_OUTPUT_DIR}")
    print(f"All audio files generated and saved to: {AUDIO_OUTPUT_DIR}")
    print(f"Created {NUM_PARTICIPANTS} participant designs with:")
    for trial_type, count in trial_counts.items():
        print(f"  - {count} {trial_type} trials")
    
    if 'estimated_duration' in validation_results:
        print(f"Estimated experiment duration: {validation_results['estimated_duration']['minutes']:.1f} minutes per participant")
    
    print("Each participant has TWO audio files: one with looming stimuli and one with tactile stimuli")
    print("Process complete.")

if __name__ == "__main__":
    main()
```

```python
"""
Optimized Peripersonal Space (PPS) Experiment Generator

This script generates PPS experiments with dynamic trial counts
and intelligently optimized audio length based on the structure of 
the box breathing audio and the specific number of trials required.
"""

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import soundfile as sf
import scipy.signal as signal
from datetime import datetime
from itertools import product, combinations
import warnings
import json
import math
import random

# ======================================================================
# CONFIGURABLE PARAMETERS - MODIFY THESE TO CUSTOMIZE YOUR EXPERIMENT
# ======================================================================

# Experiment structure
EXPERIMENT_CONFIG = {
    'repetitions_per_condition': 5,  # How many times each condition is repeated
    'included_conditions': {
        'inhalation': True,  # Include inhalation trials
        'exhalation': True,  # Include exhalation trials
        'baseline': True,    # Include baseline trials
    },
    'catch_trial_percentage': 10,  # Percentage of total trials that should be catch trials
    'debug_mode': False,     # Enable for verbose output and diagnostics
    'end_buffer_seconds': 10 # Buffer to add at end of audio (reduced from 30)
}

# SOA conditions in milliseconds (Stimulus Onset Asynchrony)
SOA_CONDITIONS_MS = [190, 400, 700, 1000, 1500]

# Jitter options in milliseconds (temporal variability)
JITTER_OPTIONS_MS = [100, 200, 300, 400, 500]

# File paths - UPDATE THESE TO MATCH YOUR SYSTEM
BASE_DIR = r"C:\Users\cogpsy-vrlab\Documents\GitHub\BreathingSpace"

# Original audio paths (update if needed)
ORIG_AUDIO_DIR = r"C:\Users\cogpsy-vrlab\Documents\PPS_module\BreathingPilot"

# Box breathing audio
BASE_AUDIO_PATH = os.path.join(ORIG_AUDIO_DIR, "AudioInstructionsBoxBreathing", "TargetStimuli", 
                              "6. Box Breathing Combined (Segment Injected).wav")

# Timestamp reference for breathing phases
TIMESTAMPS_CSV_PATH = os.path.join(ORIG_AUDIO_DIR, "AudioInstructionsBoxBreathing", "TargetStimuli", 
                                  "BoxBreathing_extended_timestamps.csv")

# Output directories
DESIGN_OUTPUT_DIR = os.path.join(BASE_DIR, "ExperimentLog")
AUDIO_OUTPUT_DIR = os.path.join(BASE_DIR, "ExperimentAudio")

# Looming stimuli (directional audio)
LOOMING_STIMULI = {
    'right': os.path.join(ORIG_AUDIO_DIR, "PPS_Experiment_Module", "right_az90_FABIAN_HRIR_natural.wav"),
    'front': os.path.join(ORIG_AUDIO_DIR, "PPS_Experiment_Module", "front_az0_FABIAN_HRIR_natural.wav"),
    'left': os.path.join(ORIG_AUDIO_DIR, "PPS_Experiment_Module", "left_az-90_FABIAN_HRIR_natural.wav")
}

# Tactile stimulus (to be synchronized with looming stimuli + SOA)
TACTILE_STIMULUS_PATH = os.path.join(ORIG_AUDIO_DIR, "PPS_Experiment_Module", "tactile_stimulus.wav")

# Number of participants to generate designs for
NUM_PARTICIPANTS = 100

# ======================================================================
# BREATHING PHASE ANALYZER
# ======================================================================

class BreathingPhaseAnalyzer:
    """
    Analyzes the timestamps CSV to identify breathing phase cycles and structure.
    This allows for more intelligent audio generation and trial distribution.
    """
    
    def __init__(self, timestamps_csv_path):
        """
        Initialize the breathing phase analyzer.
        
        Args:
            timestamps_csv_path: Path to CSV with breathing timestamps
        """
        self.timestamps_csv_path = timestamps_csv_path
        self.timestamps_df = None
        self.breathing_cycles = None
        self.phase_map = None
        
        # Load and analyze the timestamps
        self._load_timestamps()
        self._analyze_phases()
    
    def _load_timestamps(self):
        """Load timestamps from CSV file."""
        try:
            self.timestamps_df = pd.read_csv(self.timestamps_csv_path)
            
            # Convert timestamps to seconds
            if 'milliseconds' in self.timestamps_df.columns:
                self.timestamps_df['seconds'] = self.timestamps_df['milliseconds'] / 1000.0
            
            print(f"Loaded {len(self.timestamps_df)} breathing phase timestamps")
        except Exception as e:
            print(f"Error loading timestamps CSV: {e}")
            # Create a dummy timestamps structure if we can't load the real one
            self.timestamps_df = pd.DataFrame({
                'seconds': np.linspace(0, 600, 100)  # Create 100 evenly spaced timestamps over 10 minutes
            })
    
    def _analyze_phases(self):
        """
        Analyze timestamps to identify breathing cycles and phases.
        Maps each timestamp to its corresponding breathing phase.
        """
        if self.timestamps_df is None or len(self.timestamps_df) == 0:
            return
        
        # Extract relevant columns
        timestamps = self.timestamps_df.copy()
        
        # Try to identify patterns in the timestamps
        # Look for phase information in column names
        phase_columns = [col for col in timestamps.columns if 'phase' in col.lower()]
        
        if phase_columns:
            # If we have explicit phase information
            phase_col = phase_columns[0]
            timestamps['phase'] = timestamps[phase_col]
        else:
            # Otherwise, try to infer phases from the timestamp intervals
            timestamps['interval'] = timestamps['seconds'].diff()
            
            # Detect breathing pattern based on intervals between timestamps
            mean_interval = timestamps['interval'].mean()
            
            # Assign default phases if we can't detect a pattern
            timestamps['phase'] = 'unknown'
            
            phase_pattern = ['inhalation', 'hold1', 'exhalation', 'hold2']
            phase_index = 0
            
            # Assign phases to timestamps
            for i in range(len(timestamps)):
                timestamps.loc[i, 'phase'] = phase_pattern[phase_index]
                phase_index = (phase_index + 1) % len(phase_pattern)
        
        # Group timestamps by phase
        self.phase_map = {}
        for phase in timestamps['phase'].unique():
            self.phase_map[phase] = timestamps[timestamps['phase'] == phase]['seconds'].tolist()
        
        # Identify breathing cycles
        self.breathing_cycles = []
        if 'inhalation' in self.phase_map:
            # Each inhalation timestamp is the start of a new cycle
            inhalation_times = self.phase_map['inhalation']
            
            for i, inhale_time in enumerate(inhalation_times):
                cycle = {'start': inhale_time}
                if i < len(inhalation_times) - 1:
                    cycle['end'] = inhalation_times[i + 1]
                else:
                    # For the last cycle, estimate the end
                    if len(inhalation_times) > 1:
                        avg_cycle_duration = (inhalation_times[-1] - inhalation_times[0]) / (len(inhalation_times) - 1)
                        cycle['end'] = inhale_time + avg_cycle_duration
                    else:
                        cycle['end'] = inhale_time + 16  # Default to 16 second cycle if only one timestamp
                
                self.breathing_cycles.append(cycle)
        
        print(f"Identified {len(self.breathing_cycles)} breathing cycles")
        
        if EXPERIMENT_CONFIG['debug_mode'] and self.breathing_cycles:
            print(f"First cycle: {self.breathing_cycles[0]}")
            print(f"Last cycle: {self.breathing_cycles[-1]}")
            print(f"Available phases: {list(self.phase_map.keys())}")
            if 'inhalation' in self.phase_map:
                print(f"Inhalation timestamps: {len(self.phase_map['inhalation'])} timestamps")
    
    def get_phase_timestamps(self, phase_name):
        """
        Get timestamps for a specific breathing phase.
        
        Args:
            phase_name: Name of the breathing phase (e.g., 'inhalation')
            
        Returns:
            List of timestamps in seconds
        """
        if self.phase_map and phase_name in self.phase_map:
            return self.phase_map[phase_name]
        return []
    
    def get_cycle_duration(self):
        """
        Get the average duration of a complete breathing cycle.
        
        Returns:
            Average cycle duration in seconds
        """
        if not self.breathing_cycles:
            return 16.0  # Default box breathing cycle duration
        
        durations = [cycle['end'] - cycle['start'] for cycle in self.breathing_cycles]
        return sum(durations) / len(durations)
    
    def get_required_cycles(self, trial_counts):
        """
        Calculate how many complete breathing cycles are needed for the given trial counts.
        
        Args:
            trial_counts: Dictionary with trial counts for each condition
            
        Returns:
            Number of breathing cycles needed
        """
        # Calculate total number of trials
        total_trials = sum(trial_counts.values())
        
        # Get available slots per cycle for each trial type
        slots_per_cycle = {}
        for trial_type in trial_counts.keys():
            if trial_type == 'catch':
                # Catch trials can use any phase
                slots_per_cycle[trial_type] = 4  # Box breathing has 4 phases per cycle
            elif trial_type in self.phase_map:
                # Each cycle has one of each specific phase
                slots_per_cycle[trial_type] = 1
            else:
                # Default to one slot per cycle for unknown trial types
                slots_per_cycle[trial_type] = 1
        
        # Calculate cycles needed for each trial type
        cycles_needed = {}
        for trial_type, count in trial_counts.items():
            if trial_type in slots_per_cycle and slots_per_cycle[trial_type] > 0:
                cycles_needed[trial_type] = math.ceil(count / slots_per_cycle[trial_type])
            else:
                cycles_needed[trial_type] = 0
        
        # Use the maximum cycles needed for any trial type
        if cycles_needed:
            max_cycles = max(cycles_needed.values())
        else:
            max_cycles = math.ceil(total_trials / 4)  # Default assumption: 4 slots per cycle
        
        if EXPERIMENT_CONFIG['debug_mode']:
            print(f"Trial counts: {trial_counts}")
            print(f"Slots per cycle: {slots_per_cycle}")
            print(f"Cycles needed by trial type: {cycles_needed}")
            print(f"Max cycles needed: {max_cycles}")
        
        return max_cycles

# ======================================================================
# DYNAMIC TRIAL COUNT CALCULATOR
# ======================================================================

def calculate_trial_counts():
    """
    Calculate trial counts based on the experiment configuration.
    
    Returns:
        Dictionary with trial counts for each condition
    """
    # Get configuration
    reps = EXPERIMENT_CONFIG['repetitions_per_condition']
    selected_conditions = [cond for cond, include in EXPERIMENT_CONFIG['included_conditions'].items() if include]
    
    # Each condition has three stimulus types (left, front, right)
    stimuli_per_condition = 3
    
    # Calculate basic trial counts for selected conditions
    basic_trial_counts = {cond: reps * stimuli_per_condition for cond in selected_conditions}
    
    # Calculate total number of trials from selected conditions
    total_trials = sum(basic_trial_counts.values())
    
    # Calculate catch trials based on percentage
    catch_percentage = EXPERIMENT_CONFIG['catch_trial_percentage']
    total_with_catch = total_trials / (1 - catch_percentage/100)
    catch_trials = round(total_with_catch * (catch_percentage/100))
    
    # Add catch trials to the count
    final_counts = basic_trial_counts.copy()
    final_counts['catch'] = catch_trials
    
    if EXPERIMENT_CONFIG['debug_mode']:
        print(f"Trial counts calculated:")
        for cond, count in final_counts.items():
            print(f"  {cond}: {count} trials")
        print(f"  Total: {sum(final_counts.values())} trials")
    
    return final_counts

# ======================================================================
# COUNTERBALANCED DESIGN GENERATOR
# ======================================================================

class DynamicCounterbalancer:
    """
    Creates counterbalanced experimental designs with flexible trial counts
    based on configuration parameters.
    """
    
    def __init__(self, output_dir, soa_conditions, jitter_options, breathing_analyzer=None):
        """
        Initialize the counterbalancer with experiment parameters.
        
        Args:
            output_dir: Directory to save design files
            soa_conditions: List of SOA values in milliseconds
            jitter_options: List of jitter values in milliseconds
            breathing_analyzer: Optional BreathingPhaseAnalyzer instance
        """
        # Ensure output directory exists
        self.output_dir = output_dir
        os.makedirs(output_dir, exist_ok=True)
        
        # Calculate trial counts based on configuration
        self.trial_counts = calculate_trial_counts()
        
        # Define experimental factors
        self.factors = {
            'stimulus_type': ['right', 'front', 'left'],  # Stimulus directions
            'trial_type': list(self.trial_counts.keys()),  # Active trial types
            'soa_condition': list(range(len(soa_conditions))),  # SOA condition indices
            'jitter': jitter_options  # Jitter in milliseconds
        }
        
        # Store actual SOA values for reference
        self.soa_values_ms = soa_conditions
        
        # Reference to breathing analyzer for intelligent timestamp mapping
        self.breathing_analyzer = breathing_analyzer
        
        # Default timing parameters
        self.base_time = pd.Timestamp('2025-03-01 09:00:00')
        self.trial_duration = 8  # Average seconds between trials
        
        # This will be set if a reference CSV is provided
        self.reference_csv_path = None

    def generate_comprehensive_design(self, num_participants=100, reference_csv_path=None):
        """
        Generate a counterbalanced design for multiple participants.
        
        Args:
            num_participants: Number of participant designs to generate
            reference_csv_path: Path to CSV with breathing exercise timestamps
        
        Returns:
            List of DataFrames, one for each participant
        """
        self.reference_csv_path = reference_csv_path
        designs = []
        
        # Set a fixed seed for reproducibility
        np.random.seed(42)
        
        # Generate design for each participant
        for participant_id in range(num_participants):
            design = self._generate_single_participant_design(participant_id)
            designs.append(design)
        
        return designs
    
    def _generate_single_participant_design(self, participant_id):
        """
        Generate a balanced design for a single participant with precise timestamps.
        Now optimized to use available breathing phase timestamps intelligently.
        
        Args:
            participant_id: Unique identifier for the participant
        
        Returns:
            DataFrame with complete experimental design including timestamps
        """
        design = []
        
        # Get the appropriate timestamps for each trial type
        trial_type_timestamps = self._get_trial_type_timestamps()
        
        # If a reference CSV is provided, use it for timing
        if self.reference_csv_path is not None or trial_type_timestamps:
            # Generate designs for each trial type using the appropriate timestamps
            for trial_type, count in self.trial_counts.items():
                # Skip conditions with zero trials
                if count == 0:
                    continue
                
                # Get timestamps for this trial type
                available_timestamps = trial_type_timestamps.get(trial_type, [])
                
                # If no timestamps available, skip this trial type
                if not available_timestamps:
                    print(f"Warning: No timestamps available for trial type '{trial_type}'. Skipping.")
                    continue
                
                # Create a balanced stimulus sequence for this trial type
                stimulus_sequence = []
                for stim_type in self.factors['stimulus_type']:
                    # For catch trials, distribute evenly across stimulus types
                    stim_count = count // len(self.factors['stimulus_type'])
                    if stim_count > 0:
                        stimulus_sequence.extend([stim_type] * stim_count)
                
                # Handle any remainders by adding random stimuli
                remaining = count - len(stimulus_sequence)
                if remaining > 0:
                    stimulus_sequence.extend(np.random.choice(self.factors['stimulus_type'], size=remaining))
                
                # Shuffle the stimulus sequence
                np.random.shuffle(stimulus_sequence)
                
                # Ensure we have enough timestamps
                if len(available_timestamps) < len(stimulus_sequence):
                    # Repeat timestamps as needed
                    repeats_needed = math.ceil(len(stimulus_sequence) / len(available_timestamps))
                    available_timestamps = np.tile(available_timestamps, repeats_needed)[:len(stimulus_sequence)]
                
                # Shuffle timestamps to avoid sequential patterns
                shuffled_timestamps = available_timestamps.copy()
                np.random.shuffle(shuffled_timestamps)
                
                # Map timestamps to stimulus sequence
                for idx, (stim_type, timestamp_sec) in enumerate(zip(stimulus_sequence, shuffled_timestamps[:len(stimulus_sequence)])):
                    # Convert timestamp to milliseconds
                    base_ms = int(timestamp_sec * 1000)
                    
                    # Choose a random jitter value (in ms)
                    current_jitter = np.random.choice(self.factors['jitter'])
                    
                    # Get SOA condition index and actual value in milliseconds
                    soa_idx = idx % len(self.factors['soa_condition'])
                    soa_value_ms = self.soa_values_ms[soa_idx]
                    
                    # Apply jitter and SOA to get final timestamps
                    jittered_ms = base_ms + current_jitter
                    soa_ms = jittered_ms + soa_value_ms
                    
                    # Format timestamps as "mm:ss.f"
                    timestamp_original = f"{int(base_ms // 60000):02}:{(base_ms % 60000) / 1000:.1f}"
                    timestamp_after_jitter = f"{int(jittered_ms // 60000):02}:{(jittered_ms % 60000) / 1000:.1f}"
                    timestamp_with_soa = f"{int(soa_ms // 60000):02}:{(soa_ms % 60000) / 1000:.1f}"
                    
                    # Compute trial number
                    trial_number = 20 + len(design)  # Offset of 20 trials
                    
                    # For catch trials, there's no tactile component
                    is_tactile = (trial_type != 'catch')
                    
                    # Add this trial to the design
                    design.append({
                        'participant_id': participant_id,
                        'trial_number': trial_number,
                        'stimulus_type': stim_type,
                        'trial_type': trial_type,
                        'soa_condition_idx': soa_idx,
                        'soa_value_ms': soa_value_ms,  # Store exact SOA value in ms
                        'jitter_ms': current_jitter,
                        'timestamp_original': timestamp_original,
                        'timestamp_after_jitter': timestamp_after_jitter,
                        'timestamp_with_soa': timestamp_with_soa,  # New column with jitter + SOA
                        'base_ms': base_ms,
                        'jittered_ms': jittered_ms,
                        'soa_ms': soa_ms,
                        'is_tactile': is_tactile,  # Flag to indicate if trial has tactile component
                        'timestamp_sec': timestamp_sec  # Store original timestamp in seconds
                    })
                
        else:
            # If no reference CSV or timestamps available, use regular intervals
            print("Warning: No reference CSV or breathing analyzer available. Using regular intervals.")
            
            # Create a sequence of trials with regular intervals
            trial_number = 20  # Offset
            current_time_sec = 10  # Start at 10 seconds
            
            for trial_type, count in self.trial_counts.items():
                # Skip conditions with zero trials
                if count == 0:
                    continue
                
                # Create a balanced stimulus sequence for this trial type
                stimulus_sequence = []
                for stim_type in self.factors['stimulus_type']:
                    stim_count = count // len(self.factors['stimulus_type'])
                    if stim_count > 0:
                        stimulus_sequence.extend([stim_type] * stim_count)
                
                # Handle any remainders
                remaining = count - len(stimulus_sequence)
                if remaining > 0:
                    stimulus_sequence.extend(np.random.choice(self.factors['stimulus_type'], size=remaining))
                
                # Shuffle the stimulus sequence
                np.random.shuffle(stimulus_sequence)
                
                # Add each trial to the design
                for idx, stim_type in enumerate(stimulus_sequence):
                    # Choose a random jitter value (in ms)
                    current_jitter = np.random.choice(self.factors['jitter'])
                    
                    # Get SOA condition index and actual value in milliseconds
                    soa_idx = idx % len(self.factors['soa_condition'])
                    soa_value_ms = self.soa_values_ms[soa_idx]
                    
                    # Convert to milliseconds
                    base_ms = int(current_time_sec * 1000)
                    jittered_ms = base_ms + current_jitter
                    soa_ms = jittered_ms + soa_value_ms
                    
                    # Format timestamps as "mm:ss.f"
                    timestamp_original = f"{int(base_ms // 60000):02}:{(base_ms % 60000) / 1000:.1f}"
                    timestamp_after_jitter = f"{int(jittered_ms // 60000):02}:{(jittered_ms % 60000) / 1000:.1f}"
                    timestamp_with_soa = f"{int(soa_ms // 60000):02}:{(soa_ms % 60000) / 1000:.1f}"
                    
                    # For catch trials, there's no tactile component
                    is_tactile = (trial_type != 'catch')
                    
                    # Add this trial to the design
                    design.append({
                        'participant_id': participant_id,
                        'trial_number': trial_number,
                        'stimulus_type': stim_type,
                        'trial_type': trial_type,
                        'soa_condition_idx': soa_idx,
                        'soa_value_ms': soa_value_ms,
                        'jitter_ms': current_jitter,
                        'timestamp_original': timestamp_original,
                        'timestamp_after_jitter': timestamp_after_jitter,
                        'timestamp_with_soa': timestamp_with_soa,
                        'base_ms': base_ms,
                        'jittered_ms': jittered_ms,
                        'soa_ms': soa_ms,
                        'is_tactile': is_tactile,
                        'timestamp_sec': current_time_sec
                    })
                    
                    # Increment trial number and time
                    trial_number += 1
                    current_time_sec += self.trial_duration
        
        # Convert to DataFrame
        design_df = pd.DataFrame(design)
        
        # Sort by timestamp and reset index
        if not design_df.empty:
            design_df = design_df.sort_values(by='timestamp_sec').reset_index(drop=True)
        
        # Calculate and display expected experiment duration
        if EXPERIMENT_CONFIG['debug_mode'] and not design_df.empty:
            if 'timestamp_sec' in design_df.columns:
                first_time = design_df['timestamp_sec'].min()
                last_time = design_df['timestamp_sec'].max()
                duration_minutes = (last_time - first_time) / 60
                print(f"Participant {participant_id}: {len(design_df)} trials, duration ~{duration_minutes:.1f} minutes")
        
        return design_df
    
    def _get_trial_type_timestamps(self):
        """
        Get appropriate timestamps for each trial type from the breathing analyzer.
        
        Returns:
            Dictionary mapping trial types to lists of timestamps
        """
        trial_type_timestamps = {}
        
        if self.breathing_analyzer:
            # Map each trial type to appropriate breathing phase timestamps
            for trial_type in self.trial_counts.keys():
                if trial_type == 'inhalation':
                    # Use inhalation phase timestamps
                    timestamps = self.breathing_analyzer.get_phase_timestamps('inhalation')
                elif trial_type == 'exhalation':
                    # Use exhalation phase timestamps
                    timestamps = self.breathing_analyzer.get_phase_timestamps('exhalation')
                elif trial_type == 'baseline':
                    # Use hold phases (combine both holds if available)
                    hold1 = self.breathing_analyzer.get_phase_timestamps('hold1')
                    hold2 = self.breathing_analyzer.get_phase_timestamps('hold2')
                    timestamps = hold1 + hold2
                elif trial_type == 'catch':
                    # Catch trials can use any phase
                    inhalation = self.breathing_analyzer.get_phase_timestamps('inhalation')
                    exhalation = self.breathing_analyzer.get_phase_timestamps('exhalation')
                    hold1 = self.breathing_analyzer.get_phase_timestamps('hold1')
                    hold2 = self.breathing_analyzer.get_phase_timestamps('hold2')
                    timestamps = inhalation + exhalation + hold1 + hold2
                else:
                    # Unknown trial type
                    timestamps = []
                
                if timestamps:
                    trial_type_timestamps[trial_type] = sorted(timestamps)
        
        return trial_type_timestamps
    
    def save_designs(self, designs):
        """
        Save generated designs to CSV files.
        
        Args:
            designs: List of participant design DataFrames
        """
        for design in designs:
            if design.empty:
                continue
                
            participant_id = design['participant_id'].iloc[0]
            filename = f'participant_{participant_id}_design.csv'
            filepath = os.path.join(self.output_dir, filename)
            design.to_csv(filepath, index=False)
            
            if EXPERIMENT_CONFIG['debug_mode']:
                print(f"Saved design for participant {participant_id} with {len(design)} trials")
            else:
                print(f"Saved design for participant {participant_id}")
    
    def validate_designs(self, designs):
        """
        Comprehensively validate the generated designs to ensure balance.
        
        Args:
            designs: List of participant design DataFrames
        
        Returns:
            Dictionary with validation statistics
        """
        print("Starting design validation...")
        
        validation_results = {
            'stimulus_type': {},
            'trial_type': {},
            'soa_condition_idx': {},
            'jitter': {},
            'doegen_validation': {},
            'experiment_config': EXPERIMENT_CONFIG
        }
        
        try:
            # Combine all designs for overall validation
            all_designs = pd.concat(designs)
            
            # ------- Basic Distribution Validation -------
            # Validate each factor by grouping by trial type
            for factor in ['stimulus_type', 'trial_type', 'soa_condition_idx']:
                grouped_validation = all_designs.groupby(['trial_type', factor]).size() / len(designs)
                validation_results[factor] = grouped_validation.to_dict()
            
            print("Basic distribution validation complete.")
            
            # ------- Additional Dynamic Validation -------
            # Validate trial counts match configuration
            actual_counts = all_designs.groupby('trial_type').size().to_dict()
            validation_results['trial_counts'] = {
                'expected': self.trial_counts,
                'actual_per_participant': {k: v/len(designs) for k, v in actual_counts.items()}
            }
            
            # Validate SOA distribution
            soa_distribution = all_designs.groupby(['trial_type', 'soa_value_ms']).size().to_dict()
            validation_results['soa_distribution'] = soa_distribution
            
            # Find the approximate experiment duration
            if all_designs.shape[0] > 0:
                # Get last timestamp from each participant
                participant_durations = []
                for participant_id, group in all_designs.groupby('participant_id'):
                    if not group.empty:
                        if 'timestamp_sec' in group.columns:
                            # Use timestamp_sec if available
                            last_time = group['timestamp_sec'].max()
                            participant_durations.append(last_time)
                        else:
                            # Otherwise parse from timestamp string
                            last_time = group['timestamp_after_jitter'].iloc[-1]
                            parts = last_time.split(':')
                            if len(parts) == 2:
                                minutes = int(parts[0])
                                seconds = float(parts[1])
                                total_seconds = minutes * 60 + seconds
                                participant_durations.append(total_seconds)
                
                if participant_durations:
                    avg_duration = sum(participant_durations) / len(participant_durations)
                    validation_results['estimated_duration'] = {
                        'seconds': avg_duration,
                        'minutes': avg_duration / 60,
                        'per_trial': avg_duration / (sum(self.trial_counts.values()) / len(designs))
                    }
                    print(f"Estimated experiment duration: {avg_duration/60:.1f} minutes")
            
            # Save detailed validation report to JSON
            validation_report_path = os.path.join(self.output_dir, 'dynamic_validation_report.json')
            try:
                with open(validation_report_path, 'w') as f:
                    json.dump(validation_results, f, indent=4, default=str)
                print(f"Validation report saved to: {validation_report_path}")
            except Exception as e:
                print(f"Warning: Error saving validation report: {str(e)}")
            
            # Print summary of validation results
            print("\nValidation Results:")
            print(f"Total participants: {len(designs)}")
            print(f"Total trials per participant: {sum(self.trial_counts.values())}")
            print("Trial counts by condition:")
            for trial_type, count in self.trial_counts.items():
                actual = actual_counts.get(trial_type, 0) / len(designs)
                print(f"  {trial_type}: {count} expected, {actual:.1f} actual")
            
            if 'estimated_duration' in validation_results:
                print(f"Estimated duration: {validation_results['estimated_duration']['minutes']:.1f} minutes")
            
        except Exception as e:
            print(f"Error during validation: {str(e)}")
            import traceback
            traceback.print_exc()
            print("Validation could not be completed. Returning partial results.")
        
        print("Validation complete.")
        return validation_results

# ======================================================================
# OPTIMIZED AUDIO GENERATOR
# ======================================================================

class OptimizedAudioGenerator:
    """
    Generates participant-specific audio files with embedded stimuli,
    optimizing audio length based on breathing cycle analysis and trial counts.
    """
    
    def __init__(self, base_audio_path, output_dir, looming_stimuli_paths, tactile_stimulus_path, breathing_analyzer=None):
        """
        Initialize the optimized audio generator.
        
        Args:
            base_audio_path: Path to the base box breathing audio
            output_dir: Directory to save participant-specific audio files
            looming_stimuli_paths: Dictionary of paths to looming stimuli sounds
            tactile_stimulus_path: Path to the tactile stimulus sound
            breathing_analyzer: BreathingPhaseAnalyzer instance for intelligent audio generation
        """
        # Store stimulus paths
        self.looming_stimuli = looming_stimuli_paths
        self.tactile_stimulus_path = tactile_stimulus_path
        
        # Load base audio (breathing exercise)
        self.base_audio_data, self.sample_rate = sf.read(base_audio_path)
        
        # Store base audio info
        self.base_audio_duration = len(self.base_audio_data) / self.sample_rate
        
        # Breathing analyzer for cycle-aware audio optimization
        self.breathing_analyzer = breathing_analyzer
        
        if EXPERIMENT_CONFIG['debug_mode']:
            print(f"Base audio duration: {self.base_audio_duration:.1f} seconds "
                  f"({self.base_audio_duration/60:.1f} minutes)")
        
        # Load tactile stimulus
        self.tactile_data, _ = sf.read(tactile_stimulus_path)
        
        # Output directory for audio files
        self.output_dir = output_dir
        os.makedirs(output_dir, exist_ok=True)
        
        # Analyze the base audio to identify cycles/segments
        self._analyze_base_audio()
    
    def _analyze_base_audio(self):
        """
        Analyze the base audio to identify breathing cycles and segments
        that can be intelligently repeated.
        """
        # Get cycle duration from breathing analyzer if available
        self.cycle_duration = 16.0  # Default: 16 seconds for box breathing (4-4-4-4)
        
        if self.breathing_analyzer:
            self.cycle_duration = self.breathing_analyzer.get_cycle_duration()
            
            if EXPERIMENT_CONFIG['debug_mode']:
                print(f"Breathing cycle duration: {self.cycle_duration:.2f} seconds")
        
        # Calculate how many samples correspond to one breathing cycle
        self.cycle_samples = int(self.cycle_duration * self.sample_rate)
        
        # Identify how many complete cycles are in the base audio
        self.cycles_in_base = math.floor(len(self.base_audio_data) / self.cycle_samples)
        
        if EXPERIMENT_CONFIG['debug_mode']:
            print(f"Base audio contains {self.cycles_in_base} complete breathing cycles")
    
    def optimize_audio_length(self, design_df, trial_counts=None):
        """
        Calculate the optimal audio length based on the design and trial counts.
        
        Args:
            design_df: DataFrame with stimuli design
            trial_counts: Optional dictionary with trial counts for each condition
            
        Returns:
            Required audio duration in seconds
        """
        # If design is empty, return minimal duration
        if design_df.empty:
            return self.base_audio_duration
        
        # Method 1: Use the latest timestamp from the design
        if 'timestamp_sec' in design_df.columns:
            latest_time_sec = design_df['timestamp_sec'].max()
        else:
            # Parse from timestamp string
            latest_timestamp = design_df['timestamp_with_soa'].iloc[-1]
            latest_time_parts = latest_timestamp.split(':')
            minutes = int(latest_time_parts[0])
            seconds = float(latest_time_parts[1])
            latest_time_sec = minutes * 60 + seconds
        
        # Method 2: Use breathing analyzer to calculate required cycles
        if self.breathing_analyzer and trial_counts:
            # Calculate cycles needed based on trial distribution
            required_cycles = self.breathing_analyzer.get_required_cycles(trial_counts)
            required_from_cycles = required_cycles * self.cycle_duration
            
            # Use the longer of the two estimates
            required_duration = max(latest_time_sec, required_from_cycles)
        else:
            # Add buffer to latest timestamp
            required_duration = latest_time_sec
        
        # Add end buffer for concluding instructions
        required_duration += EXPERIMENT_CONFIG['end_buffer_seconds']
        
        if EXPERIMENT_CONFIG['debug_mode']:
            print(f"Optimized audio duration: {required_duration:.1f} seconds "
                  f"({required_duration/60:.1f} minutes)")
        
        return required_duration
    
    def generate_optimized_audio(self, required_duration):
        """
        Generate optimized audio of the specified duration by intelligently
        repeating breathing cycles from the base audio.
        
        Args:
            required_duration: Required duration in seconds
            
        Returns:
            Optimized audio data
        """
        # Calculate required samples
        required_samples = int(required_duration * self.sample_rate)
        
        # If required duration is less than base audio, simply trim
        if required_samples <= len(self.base_audio_data):
            return self.base_audio_data[:required_samples]
        
        # Calculate how many complete cycles we need
        required_cycles = math.ceil(required_duration / self.cycle_duration)
        
        if EXPERIMENT_CONFIG['debug_mode']:
            print(f"Generating optimized audio: need {required_cycles} cycles "
                  f"({required_duration:.1f} seconds)")
        
        # Check if we have enough cycles in the base audio
        if required_cycles <= self.cycles_in_base:
            # We can just use the base audio with appropriate trimming
            return self.base_audio_data[:required_samples]
        
        # We need to create an extended version with repeated cycles
        # Identify the middle cycles to repeat (avoid first and last if possible)
        if self.cycles_in_base >= 3:
            # Extract a middle cycle that we can repeat
            cycle_start = self.cycle_samples  # Skip first cycle
            cycle_end = cycle_start + self.cycle_samples
            repeatable_cycle = self.base_audio_data[cycle_start:cycle_end]
        else:
            # Extract whatever complete cycle we have
            repeatable_cycle = self.base_audio_data[:self.cycle_samples]
        
        # Create the extended audio
        # Start with the original base audio
        extended_audio = self.base_audio_data.copy()
        
        # Add repeatable cycles until we reach the required duration
        while len(extended_audio) < required_samples:
            extended_audio = np.vstack((extended_audio, repeatable_cycle))
        
        # Trim to exact required length
        return extended_audio[:required_samples]
    
    def generate_participant_audio(self, participant_id, design_df, trial_counts=None):
        """
        Generate two synchronized audio files for a participant with optimized length.
        
        Args:
            participant_id: Participant identifier
            design_df: DataFrame with stimuli design
            trial_counts: Optional dictionary with trial counts
                
        Returns:
            Tuple (looming_audio, tactile_audio, injection_log)
        """
        # Check if design is empty
        if design_df.empty:
            print(f"Warning: Empty design for participant {participant_id}. Skipping audio generation.")
            return None, None, pd.DataFrame()
        
        # Calculate optimal audio length
        required_duration = self.optimize_audio_length(design_df, trial_counts)
        
        # Generate optimized audio
        optimized_base_audio = self.generate_optimized_audio(required_duration)
        
        # Create two separate audio streams
        looming_audio = optimized_base_audio.copy()  # Box breathing with looming sounds
        tactile_audio = np.zeros_like(looming_audio)  # Empty track matching looming shape
        
        # Ensure tactile stimulus is compatible format
        if self.tactile_data.ndim == 1 and looming_audio.ndim > 1:  # If tactile is mono and looming is stereo
            tactile_stereo = np.column_stack((self.tactile_data, self.tactile_data))
        elif self.tactile_data.ndim > 1 and looming_audio.ndim == 1:  # If tactile is stereo and looming is mono
            tactile_stereo = np.mean(self.tactile_data, axis=1)
        else:
            tactile_stereo = self.tactile_data
            
        # Initialize log to track all stimulus injections
        injection_log = []
        
        # Verify required columns
        required_columns = ['stimulus_type', 'jittered_ms', 'soa_ms', 'soa_value_ms']
        missing_columns = [col for col in required_columns if col not in design_df.columns]
        if missing_columns:
            print(f"Warning: Design DataFrame missing required columns: {missing_columns}. "
                  f"Cannot ensure precise synchronization.")
            return None, None, pd.DataFrame()
        
        # Process each trial in the design
        for _, row in design_df.iterrows():
            # Get exact millisecond timestamps from the design CSV
            jittered_ms = row['jittered_ms']  # For looming stimuli
            soa_ms = row['soa_ms']  # For tactile stimuli
            
            # Convert to sample indices
            looming_injection_point = int(jittered_ms / 1000 * self.sample_rate)
            tactile_injection_point = int(soa_ms / 1000 * self.sample_rate)
            
            # Get SOA value for logging
            soa_value_ms = row['soa_value_ms']
            
            # Load the appropriate looming stimulus based on design
            stim_type = row['stimulus_type']
            looming_path = self.looming_stimuli.get(stim_type)
            if looming_path is None:
                continue  # Skip if stimulus type not found
            
            # Read the looming stimulus audio file
            looming_data, _ = sf.read(looming_path)
            
            # Check if both stimuli fit within the audio bounds
            looming_fits = (looming_injection_point + len(looming_data) <= len(looming_audio))
            
            if looming_fits:
                # Inject looming stimulus into the first audio track
                try:
                    looming_audio[looming_injection_point:looming_injection_point+len(looming_data)] += looming_data
                except ValueError as e:
                    print(f"Error injecting looming stimulus: {e}, shapes: "
                          f"looming_audio={looming_audio.shape}, looming_data={looming_data.shape}")
                    continue
                
                # Inject tactile stimulus into the second audio track for non-catch trials
                if row.get('is_tactile', True) and row['trial_type'] != 'catch':
                    tactile_fits = (tactile_injection_point >= 0 and 
                                  tactile_injection_point + len(tactile_stereo) <= len(tactile_audio))
                    
                    if tactile_fits:
                        try:
                            tactile_audio[tactile_injection_point:tactile_injection_point+len(tactile_stereo)] += tactile_stereo
                        except ValueError as e:
                            print(f"Error injecting tactile stimulus: {e}, shapes: "
                                  f"tactile_audio={tactile_audio.shape}, tactile_stereo={tactile_stereo.shape}")
                            continue
                    
                # Log details of this injection
                injection_log.append({
                    'participant_id': participant_id,
                    'trial_number': row.get('trial_number', 0),
                    'stimulus_type': stim_type,
                    'soa_condition_idx': row.get('soa_condition_idx', 0),
                    'soa_value_ms': soa_value_ms,
                    'looming_injection_sample': looming_injection_point,
                    'tactile_injection_sample': tactile_injection_point,
                    'looming_time_ms': looming_injection_point / self.sample_rate * 1000,
                    'tactile_time_ms': tactile_injection_point / self.sample_rate * 1000,
                    'jittered_ms': jittered_ms,
                    'soa_ms': soa_ms,
                    'is_tactile': row.get('is_tactile', True),
                    'trial_type': row.get('trial_type', 'unknown')
                })
        
        return looming_audio, tactile_audio, pd.DataFrame(injection_log)
    
    def save_participant_files(self, participant_id, looming_audio, tactile_audio, injection_log, design_filename):
        """
        Save participant-specific audio and log files.
        
        Args:
            participant_id: Participant identifier
            looming_audio: Modified audio with looming stimuli
            tactile_audio: Audio track with tactile stimuli
            injection_log: DataFrame with injection details
            design_filename: Base design filename for naming consistency
        
        Returns:
            Tuple of saved file paths
        """
        # Check if audio data is None
        if looming_audio is None or tactile_audio is None:
            print(f"Warning: Cannot save audio for participant {participant_id} - missing audio data")
            return None, None, None
        
        # Use the design CSV base name (remove .csv extension)
        base_name = os.path.splitext(design_filename)[0]
        
        # Define filenames for output files
        looming_filename = f"{base_name}_looming.wav"
        tactile_filename = f"{base_name}_tactile.wav"
        log_filename = f"{base_name}_stimuli_log.csv"
        
        # Create full paths
        looming_path = os.path.join(self.output_dir, looming_filename)
        tactile_path = os.path.join(self.output_dir, tactile_filename)
        log_path = os.path.join(self.output_dir, log_filename)
        
        # Save the audio files
        sf.write(looming_path, looming_audio, self.sample_rate)
        sf.write(tactile_path, tactile_audio, self.sample_rate)
        
        # Save the detailed injection log (for analysis)
        if not injection_log.empty:
            injection_log.to_csv(log_path, index=False)
        
        # Log file sizes
        looming_size_mb = os.path.getsize(looming_path) / (1024 * 1024)
        tactile_size_mb = os.path.getsize(tactile_path) / (1024 * 1024)
        
        print(f"Looming audio saved: {looming_filename} ({looming_size_mb:.1f} MB)")
        print(f"Tactile audio saved: {tactile_filename} ({tactile_size_mb:.1f} MB)")
        print(f"Injection log saved: {log_filename}")
        
        # Report audio duration
        audio_duration = len(looming_audio) / self.sample_rate
        print(f"Audio duration: {audio_duration:.1f} seconds ({audio_duration/60:.1f} minutes)")
        
        return looming_path, tactile_path, log_path


# ======================================================================
# MAIN FUNCTION
# ======================================================================

def main():
    """
    Main function to run the optimized PPS experiment setup process.
    """
    # Ensure output directories exist
    os.makedirs(DESIGN_OUTPUT_DIR, exist_ok=True)
    os.makedirs(AUDIO_OUTPUT_DIR, exist_ok=True)
    
    # Initialize breathing phase analyzer
    print("\n===== INITIALIZING BREATHING PHASE ANALYZER =====")
    breathing_analyzer = BreathingPhaseAnalyzer(TIMESTAMPS_CSV_PATH)
    
    # Calculate and display trial counts
    trial_counts = calculate_trial_counts()
    total_trials = sum(trial_counts.values())
    
    # Display configuration overview
    print("\n===== OPTIMIZED PPS EXPERIMENT GENERATOR =====")
    print(f"Repetitions per condition: {EXPERIMENT_CONFIG['repetitions_per_condition']}")
    
    # Show which conditions are included
    active_conditions = [cond for cond, include in EXPERIMENT_CONFIG['included_conditions'].items() if include]
    print("Included conditions:", ", ".join(active_conditions))
    
    print(f"Catch trial percentage: {EXPERIMENT_CONFIG['catch_trial_percentage']}%")
    print(f"Total trials per participant: {total_trials}")
    print(f"Output directory: {DESIGN_OUTPUT_DIR}")
    
    # Calculate expected cycles needed
    expected_cycles = breathing_analyzer.get_required_cycles(trial_counts)
    cycle_duration = breathing_analyzer.get_cycle_duration()
    expected_duration = expected_cycles * cycle_duration
    
    print(f"Estimated breathing cycles needed: {expected_cycles}")
    print(f"Estimated audio duration: {expected_duration:.1f} seconds ({expected_duration/60:.1f} minutes)")
    
    # Step 1: Generate counterbalanced designs
    print("\n===== STEP 1: GENERATING COUNTERBALANCED DESIGNS =====")
    print(f"Using SOA conditions (ms): {SOA_CONDITIONS_MS}")
    
    counterbalancer = DynamicCounterbalancer(
        output_dir=DESIGN_OUTPUT_DIR,
        soa_conditions=SOA_CONDITIONS_MS,
        jitter_options=JITTER_OPTIONS_MS,
        breathing_analyzer=breathing_analyzer
    )
    
    designs = counterbalancer.generate_comprehensive_design(
        num_participants=NUM_PARTICIPANTS, 
        reference_csv_path=TIMESTAMPS_CSV_PATH
    )
    
    # Save designs to CSV files
    counterbalancer.save_designs(designs)
    
    # Run validation checks
    print("\n===== RUNNING VALIDATION CHECKS =====")
    validation_results = counterbalancer.validate_designs(designs)
    
    # Step 2: Generate participant-specific audio files
    print("\n===== STEP 2: GENERATING OPTIMIZED AUDIO FILES =====")
    print(f"Creating TWO audio files per participant with optimized length:")
    print(f"  - Looming audio: Box breathing with directional looming sounds")
    print(f"  - Tactile audio: Synchronized tactile stimuli with SOA offsets")
    
    # Initialize the optimized audio generator
    audio_generator = OptimizedAudioGenerator(
        base_audio_path=BASE_AUDIO_PATH,
        output_dir=AUDIO_OUTPUT_DIR,
        looming_stimuli_paths=LOOMING_STIMULI,
        tactile_stimulus_path=TACTILE_STIMULUS_PATH,
        breathing_analyzer=breathing_analyzer
    )
    
    # Process each design and generate audio pairs
    for i, design_df in enumerate(designs):
        if design_df.empty:
            continue
            
        participant_id = design_df['participant_id'].iloc[0]
        design_filename = f"participant_{participant_id}_design.csv"
        
        print(f"\nProcessing participant {participant_id} ({i+1}/{NUM_PARTICIPANTS})...")
        
        # Verify design CSV has necessary columns
        if 'jittered_ms' not in design_df.columns or 'soa_ms' not in design_df.columns:
            print(f"Warning: Design for participant {participant_id} missing critical timing columns.")
            print("Reading design from saved CSV file to ensure correct timing...")
            
            # Re-read the design CSV
            design_file_path = os.path.join(DESIGN_OUTPUT_DIR, design_filename)
            if os.path.exists(design_file_path):
                design_df = pd.read_csv(design_file_path)
                print("Successfully loaded design from CSV with all timing information.")
            else:
                print(f"Error: Cannot find design file {design_file_path}")
                continue
        
        # Generate audio for this participant
        try:
            looming_audio, tactile_audio, injection_log = audio_generator.generate_participant_audio(
                participant_id, 
                design_df,
                trial_counts
            )
            
            # Save both audio files and the log
            audio_generator.save_participant_files(
                participant_id, 
                looming_audio, 
                tactile_audio, 
                injection_log, 
                design_filename
            )
            
            # Verify synchronization
            print("Verifying synchronization...")
            if not injection_log.empty:
                for i, row in injection_log.head(3).iterrows():
                    looming_time = row['looming_time_ms']
                    tactile_time = row['tactile_time_ms']
                    expected_diff = row['soa_value_ms']
                    actual_diff = tactile_time - looming_time
                    
                    print(f"  Trial {i+1}: SOA expected={expected_diff}ms, actual={actual_diff:.2f}ms, " +
                          f"difference={abs(actual_diff-expected_diff):.4f}ms")
                
                if len(injection_log) > 3:
                    print(f"  ... and {len(injection_log) - 3} more trials")
            else:
                print("  Warning: No stimuli were injected - check design parameters")
                
        except Exception as e:
            print(f"Error generating audio for participant {participant_id}: {str(e)}")
            import traceback
            traceback.print_exc()
            continue
    
    # Print summary
    print(f"\n===== GENERATION COMPLETE =====")
    print(f"All designs generated and saved to: {DESIGN_OUTPUT_DIR}")
    print(f"All audio files generated and saved to: {AUDIO_OUTPUT_DIR}")
    print(f"Created {NUM_PARTICIPANTS} participant designs with:")
    for trial_type, count in trial_counts.items():
        print(f"  - {count} {trial_type} trials")
    
    if 'estimated_duration' in validation_results:
        print(f"Estimated experiment duration: {validation_results['estimated_duration']['minutes']:.1f} minutes per participant")
    
    print("Each participant has TWO audio files: one with looming stimuli and one with tactile stimuli")
    print("Process complete.")

if __name__ == "__main__":
    main()
```

```python
"""
Peripersonal Space (PPS) Experiment Generator with Dynamic Audio

This script generates customizable PPS experiments with automatic 
audio adjustment based on the exact number of trials needed.

The script takes the CSV timestamps as reference and creates
a custom audio file by looping only the needed segments.
"""

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import soundfile as sf
import scipy.signal as signal
from datetime import datetime
from itertools import product, combinations
import warnings
import json
import math
import random

# ======================================================================
# CONFIGURABLE PARAMETERS - MODIFY THESE TO CUSTOMIZE YOUR EXPERIMENT
# ======================================================================

# Experiment structure
EXPERIMENT_CONFIG = {
    'repetitions_per_condition': 1,  # How many times each condition is repeated (1=minimal)
    'included_conditions': {
        'inhalation': True,  # Include inhalation trials
        'exhalation': True,  # Include exhalation trials
        'baseline': True,    # Include baseline trials
    },
    'catch_trial_percentage': 10,  # Percentage of total trials that should be catch trials
    'debug_mode': True,      # Enable for verbose output and diagnostics
}

# SOA conditions in milliseconds (Stimulus Onset Asynchrony)
SOA_CONDITIONS_MS = [190, 400, 700, 1000, 1500]

# Jitter options in milliseconds (temporal variability)
JITTER_OPTIONS_MS = [100, 200, 300, 400, 500]

# File paths - UPDATE THESE TO MATCH YOUR SYSTEM
BASE_DIR = r"C:\Users\cogpsy-vrlab\Documents\GitHub\BreathingSpace"

# Original audio paths (update if needed)
ORIG_AUDIO_DIR = r"C:\Users\cogpsy-vrlab\Documents\PPS_module\BreathingPilot"

# Box breathing audio
BASE_AUDIO_PATH = os.path.join(ORIG_AUDIO_DIR, "AudioInstructionsBoxBreathing", "TargetStimuli", 
                              "6. Box Breathing Combined (Segment Injected).wav")

# Timestamp reference for breathing phases
TIMESTAMPS_CSV_PATH = os.path.join(ORIG_AUDIO_DIR, "AudioInstructionsBoxBreathing", "TargetStimuli", 
                                  "BoxBreathing_extended_timestamps.csv")

# Output directories
DESIGN_OUTPUT_DIR = os.path.join(BASE_DIR, "ExperimentLog")
AUDIO_OUTPUT_DIR = os.path.join(BASE_DIR, "ExperimentAudio")

# Looming stimuli (directional audio)
LOOMING_STIMULI = {
    'right': os.path.join(ORIG_AUDIO_DIR, "PPS_Experiment_Module", "right_az90_FABIAN_HRIR_natural.wav"),
    'front': os.path.join(ORIG_AUDIO_DIR, "PPS_Experiment_Module", "front_az0_FABIAN_HRIR_natural.wav"),
    'left': os.path.join(ORIG_AUDIO_DIR, "PPS_Experiment_Module", "left_az-90_FABIAN_HRIR_natural.wav")
}

# Tactile stimulus (to be synchronized with looming stimuli + SOA)
TACTILE_STIMULUS_PATH = os.path.join(ORIG_AUDIO_DIR, "PPS_Experiment_Module", "tactile_stimulus.wav")

# Number of participants to generate designs for (set to 2 for testing)
NUM_PARTICIPANTS = 2  # Set to a lower number for testing; change to desired value for production

# ======================================================================
# DYNAMIC TRIAL COUNT CALCULATOR
# ======================================================================

def calculate_trial_counts():
    """
    Calculate trial counts based on the experiment configuration.
    
    Returns:
        Dictionary with trial counts for each condition
    """
    # Get configuration
    reps = EXPERIMENT_CONFIG['repetitions_per_condition']
    selected_conditions = [cond for cond, include in EXPERIMENT_CONFIG['included_conditions'].items() if include]
    
    # Each condition has three stimulus types (left, front, right)
    stimuli_per_condition = 3
    
    # Calculate basic trial counts for selected conditions
    basic_trial_counts = {cond: reps * stimuli_per_condition for cond in selected_conditions}
    
    # Calculate total number of trials from selected conditions
    total_trials = sum(basic_trial_counts.values())
    
    # Calculate catch trials based on percentage
    catch_percentage = EXPERIMENT_CONFIG['catch_trial_percentage']
    total_with_catch = total_trials / (1 - catch_percentage/100)
    catch_trials = round(total_with_catch * (catch_percentage/100))
    
    # Add catch trials to the count
    final_counts = basic_trial_counts.copy()
    final_counts['catch'] = catch_trials
    
    if EXPERIMENT_CONFIG['debug_mode']:
        print(f"Trial counts calculated:")
        for cond, count in final_counts.items():
            print(f"  {cond}: {count} trials")
        print(f"  Total: {sum(final_counts.values())} trials")
    
    return final_counts

# ======================================================================
# BREATHING CYCLE ANALYZER
# ======================================================================

class BreathingCycleAnalyzer:
    """
    Analyzes the breathing cycle timestamps to identify repeatable segments
    and calculate total cycles needed.
    """
    
    def __init__(self, timestamps_csv_path):
        """Initialize the analyzer with the CSV path"""
        self.timestamps_path = timestamps_csv_path
        self.timestamps_df = None
        self.start_timestamps = []
        self.cycle_durations = []
        self.cycle_duration_ms = 0
        self.total_cycles = 0
        
        # Load and analyze timestamps
        self._load_timestamps()
        self._analyze_cycles()
    
    def _load_timestamps(self):
        """Load timestamps from CSV file"""
        try:
            self.timestamps_df = pd.read_csv(self.timestamps_path)
            if EXPERIMENT_CONFIG['debug_mode']:
                print(f"Loaded {len(self.timestamps_df)} timestamps from CSV")
        except Exception as e:
            print(f"Error loading timestamps: {e}")
            raise
    
    def _analyze_cycles(self):
        """
        Analyze timestamps to identify breathing cycles.
        Each cycle consists of:
        - Inhale
        - Hold (after inhale)
        - Exhale
        - Hold (after exhale)
        """
        if self.timestamps_df is None or self.timestamps_df.empty:
            return
        
        # Extract timestamps
        timestamps = self.timestamps_df['milliseconds'].values
        
        # Calculate cycle information (using first few timestamps as sample)
        if len(timestamps) >= 4:
            # Calculate the duration of a single cycle
            cycle_ms = timestamps[4] - timestamps[0]
            self.cycle_duration_ms = cycle_ms
            
            if EXPERIMENT_CONFIG['debug_mode']:
                print(f"Detected cycle duration: {cycle_ms/1000:.2f} seconds")
            
            # Identify all cycle start points
            self.start_timestamps = [ts for i, ts in enumerate(timestamps) if i % 4 == 0]
            
            # Calculate durations between consecutive cycle starts
            if len(self.start_timestamps) > 1:
                self.cycle_durations = [self.start_timestamps[i+1] - self.start_timestamps[i] 
                                       for i in range(len(self.start_timestamps)-1)]
    
    def get_cycles_for_trials(self, num_trials):
        """
        Calculate how many breathing cycles are needed for the given number of trials.
        
        Args:
            num_trials: Number of trials to accommodate
            
        Returns:
            Number of breathing cycles needed
        """
        if not self.timestamps_df.empty:
            # Each breathing cycle can accommodate one trial
            # Adding a small buffer of 1 cycle
            cycles_needed = num_trials + 1
            self.total_cycles = cycles_needed
            return cycles_needed
        return 0
    
    def get_timestamps_for_cycles(self, num_cycles):
        """
        Get the timestamps needed for the specified number of cycles.
        
        Args:
            num_cycles: Number of breathing cycles
            
        Returns:
            Array of timestamps in milliseconds
        """
        if self.timestamps_df is None or self.timestamps_df.empty:
            return []
        
        # Find how many complete cycles we have in the original data
        complete_cycles = len(self.start_timestamps)
        
        if EXPERIMENT_CONFIG['debug_mode']:
            print(f"Complete cycles in original data: {complete_cycles}")
            print(f"Cycles needed: {num_cycles}")
        
        if num_cycles <= complete_cycles:
            # We have enough cycles in the original data
            return self.timestamps_df['milliseconds'].values[:num_cycles*4]
        else:
            # We need to extend the timestamps
            orig_timestamps = self.timestamps_df['milliseconds'].values
            
            # Use the average cycle duration to extend
            if self.cycle_durations:
                avg_cycle_ms = sum(self.cycle_durations) / len(self.cycle_durations)
            else:
                avg_cycle_ms = self.cycle_duration_ms
            
            # Start with original timestamps
            extended_timestamps = list(orig_timestamps.copy())
            
            # Get the last timestamp
            last_ts = orig_timestamps[-1]
            
            # Extend by adding cycles
            cycles_to_add = num_cycles - complete_cycles
            
            for c in range(cycles_to_add):
                # Add one cycle worth of timestamps (4 timestamps per cycle)
                for i in range(4):
                    # Original offset within a cycle
                    offset = orig_timestamps[i % 4] - orig_timestamps[0]
                    
                    # Calculate new timestamp by adding:
                    # 1. Last timestamp in original data
                    # 2. Current cycle offset (c+1 because we're starting after the original data)
                    # 3. Offset within the cycle
                    new_ts = last_ts + (c+1) * avg_cycle_ms + offset
                    extended_timestamps.append(new_ts)
            
            return np.array(extended_timestamps)

# ======================================================================
# COUNTERBALANCED DESIGN GENERATOR
# ======================================================================

class DynamicCounterbalancer:
    """
    Creates counterbalanced experimental designs with flexible trial counts
    based on configuration parameters.
    """
    
    def __init__(self, output_dir, soa_conditions, jitter_options):
        """
        Initialize the counterbalancer with experiment parameters.
        
        Args:
            output_dir: Directory to save design files
            soa_conditions: List of SOA values in milliseconds
            jitter_options: List of jitter values in milliseconds
        """
        # Ensure output directory exists
        self.output_dir = output_dir
        os.makedirs(output_dir, exist_ok=True)
        
        # Calculate trial counts based on configuration
        self.trial_counts = calculate_trial_counts()
        
        # Define experimental factors
        self.factors = {
            'stimulus_type': ['right', 'front', 'left'],  # Stimulus directions
            'trial_type': list(self.trial_counts.keys()),  # Active trial types
            'soa_condition': list(range(len(soa_conditions))),  # SOA condition indices
            'jitter': jitter_options  # Jitter in milliseconds
        }
        
        # Store actual SOA values for reference
        self.soa_values_ms = soa_conditions
        
        # Default timing parameters
        self.base_time = pd.Timestamp('2025-03-01 09:00:00')
        self.trial_duration = 8  # Average seconds between trials
        
        # This will be set if a reference CSV is provided
        self.reference_csv_path = None
        self.breathing_analyzer = None

    def generate_comprehensive_design(self, num_participants=100, reference_csv_path=None):
        """
        Generate a counterbalanced design for multiple participants.
        
        Args:
            num_participants: Number of participant designs to generate
            reference_csv_path: Path to CSV with breathing exercise timestamps
        
        Returns:
            List of DataFrames, one for each participant
        """
        self.reference_csv_path = reference_csv_path
        
        # Initialize the breathing cycle analyzer if we have a CSV
        if reference_csv_path:
            self.breathing_analyzer = BreathingCycleAnalyzer(reference_csv_path)
        
        # Set a fixed seed for reproducibility
        np.random.seed(42)
        
        # Generate design for each participant
        designs = []
        for participant_id in range(num_participants):
            design = self._generate_single_participant_design(participant_id)
            designs.append(design)
        
        return designs
    
    def _generate_single_participant_design(self, participant_id):
        """
        Generate a balanced design for a single participant with precise timestamps.
        
        Args:
            participant_id: Unique identifier for the participant
        
        Returns:
            DataFrame with complete experimental design including timestamps
        """
        design = []
        
        # Calculate total trials
        total_trials = sum(self.trial_counts.values())
        
        # Get timestamps based on breathing cycles if analyzer is available
        if self.breathing_analyzer:
            # Calculate needed cycles
            cycles_needed = self.breathing_analyzer.get_cycles_for_trials(total_trials)
            
            # Get extended timestamps if needed
            csv_milliseconds = self.breathing_analyzer.get_timestamps_for_cycles(cycles_needed)
            
            if EXPERIMENT_CONFIG['debug_mode']:
                print(f"Using {len(csv_milliseconds)} timestamps for {total_trials} trials")
            
            # Initialize CSV index
            csv_index = 0
        else:
            # If no analyzer, use evenly spaced timestamps
            csv_milliseconds = np.array([i * 8000 for i in range(total_trials * 2)])
            csv_index = 0
        
        # Generate designs for each trial type (inhalation, exhalation, etc.)
        for trial_type, count in self.trial_counts.items():
            # Skip conditions with zero trials
            if count == 0:
                continue
                
            # Create a balanced stimulus sequence for each trial type
            stimulus_sequence = []
            for stim_type in self.factors['stimulus_type']:
                # For catch trials, distribute evenly across stimulus types
                stim_count = count // len(self.factors['stimulus_type'])
                if stim_count > 0:
                    stimulus_sequence.extend([stim_type] * stim_count)
            
            # Handle any remainders by adding random stimuli
            remaining = count - len(stimulus_sequence)
            if remaining > 0:
                stimulus_sequence.extend(np.random.choice(self.factors['stimulus_type'], size=remaining))
            
            # Shuffle the stimulus sequence
            np.random.shuffle(stimulus_sequence)
            
            # Add design details for each trial
            for idx, stimulus_type in enumerate(stimulus_sequence):
                # Compute trial number
                trial_idx = len(design)
                trial_number = 20 + trial_idx  # Offset of 20 trials
                
                # Choose a random jitter value (in ms)
                current_jitter = np.random.choice(self.factors['jitter'])
                
                # Get SOA condition index and actual value in milliseconds
                # For catch trials, SOA isn't meaningful but we assign a value for consistency
                soa_idx = idx % len(self.factors['soa_condition'])
                soa_value_ms = self.soa_values_ms[soa_idx]
                
                # Generate timestamps (either from CSV or calculated)
                if len(csv_milliseconds) > 0:
                    # Use timestamp from the list of milliseconds
                    base_ms = csv_milliseconds[csv_index % len(csv_milliseconds)]
                    csv_index += 1
                    jittered_ms = base_ms + current_jitter
                    soa_ms = jittered_ms + soa_value_ms
                    
                    # Format timestamps as "mm:ss.f"
                    timestamp_original = f"{int(base_ms // 60000):02}:{(base_ms % 60000) / 1000:.1f}"
                    timestamp_after_jitter = f"{int(jittered_ms // 60000):02}:{(jittered_ms % 60000) / 1000:.1f}"
                    timestamp_with_soa = f"{int(soa_ms // 60000):02}:{(soa_ms % 60000) / 1000:.1f}"
                    
                    # Store raw millisecond values for precise audio syncing
                    base_ms_raw = base_ms
                    jittered_ms_raw = jittered_ms
                    soa_ms_raw = soa_ms
                else:
                    # Use base_time and trial_duration method (regular intervals)
                    t_orig = self.base_time + pd.Timedelta(seconds=trial_number * self.trial_duration)
                    t_jitter = self.base_time + pd.Timedelta(seconds=(trial_number * self.trial_duration) + (current_jitter / 1000))
                    t_soa = self.base_time + pd.Timedelta(seconds=(trial_number * self.trial_duration) + (current_jitter / 1000) + (soa_value_ms / 1000))
                    
                    timestamp_original = t_orig.strftime('%M:%S.%f')[:-4]
                    timestamp_after_jitter = t_jitter.strftime('%M:%S.%f')[:-4]
                    timestamp_with_soa = t_soa.strftime('%M:%S.%f')[:-4]
                    
                    # Calculate raw millisecond values for precise audio syncing
                    base_ms_raw = trial_number * self.trial_duration * 1000
                    jittered_ms_raw = base_ms_raw + current_jitter
                    soa_ms_raw = jittered_ms_raw + soa_value_ms
                
                # For catch trials, there's no tactile component, so set is_tactile flag
                is_tactile = (trial_type != 'catch')
                
                # Add this trial to the design
                design.append({
                    'participant_id': participant_id,
                    'trial_number': trial_number,
                    'stimulus_type': stimulus_type,
                    'trial_type': trial_type,
                    'soa_condition_idx': soa_idx,
                    'soa_value_ms': soa_value_ms,  # Store exact SOA value in ms
                    'jitter_ms': current_jitter,
                    'timestamp_original': timestamp_original,
                    'timestamp_after_jitter': timestamp_after_jitter,
                    'timestamp_with_soa': timestamp_with_soa,  # New column with jitter + SOA
                    'base_ms': base_ms_raw,
                    'jittered_ms': jittered_ms_raw,
                    'soa_ms': soa_ms_raw,
                    'is_tactile': is_tactile  # Flag to indicate if trial has tactile component
                })
        
        # Convert to DataFrame
        design_df = pd.DataFrame(design)
        
        # Shuffle within each trial type to maintain balance
        shuffled_design = []
        for trial_type in self.trial_counts.keys():
            type_subset = design_df[design_df['trial_type'] == trial_type].copy()
            if not type_subset.empty:
                type_subset = type_subset.sample(frac=1, random_state=participant_id).reset_index(drop=True)
                shuffled_design.append(type_subset)
        
        # Recombine, sort by timestamp, and reset index
        design_df = pd.concat(shuffled_design).reset_index(drop=True)
        design_df = design_df.sort_values(by='timestamp_after_jitter').reset_index(drop=True)
        
        # Calculate and display expected experiment duration
        if EXPERIMENT_CONFIG['debug_mode'] and len(design_df) > 0:
            first_time = self._parse_timestamp(design_df['timestamp_original'].iloc[0])
            last_time = self._parse_timestamp(design_df['timestamp_original'].iloc[-1])
            duration_minutes = (last_time - first_time) / 60
            print(f"Participant {participant_id}: {len(design_df)} trials, duration ~{duration_minutes:.1f} minutes")
        
        return design_df
    
    def _parse_timestamp(self, timestamp_str):
        """Parse mm:ss.f timestamp to seconds"""
        parts = timestamp_str.split(':')
        if len(parts) == 2:
            minutes = int(parts[0])
            seconds = float(parts[1])
            return minutes * 60 + seconds
        return 0
    
    def save_designs(self, designs):
        """
        Save generated designs to CSV files.
        
        Args:
            designs: List of participant design DataFrames
        """
        for design in designs:
            if design.empty:
                continue
                
            participant_id = design['participant_id'].iloc[0]
            filename = f'participant_{participant_id}_design.csv'
            filepath = os.path.join(self.output_dir, filename)
            design.to_csv(filepath, index=False)
            
            if EXPERIMENT_CONFIG['debug_mode']:
                print(f"Saved design for participant {participant_id} with {len(design)} trials")
            else:
                print(f"Saved design for participant {participant_id}")
    
    def validate_designs(self, designs):
        """
        Comprehensively validate the generated designs to ensure balance.
        
        Args:
            designs: List of participant design DataFrames
        
        Returns:
            Dictionary with validation statistics
        """
        print("Starting design validation...")
        
        validation_results = {
            'stimulus_type': {},
            'trial_type': {},
            'soa_condition_idx': {},
            'jitter': {},
            'doegen_validation': {},
            'experiment_config': EXPERIMENT_CONFIG
        }
        
        try:
            # Combine all designs for overall validation
            all_designs = pd.concat(designs)
            
            # ------- Basic Distribution Validation -------
            # Validate each factor by grouping by trial type
            for factor in ['stimulus_type', 'trial_type', 'soa_condition_idx']:
                grouped_validation = all_designs.groupby(['trial_type', factor]).size() / len(designs)
                validation_results[factor] = grouped_validation.to_dict()
            
            print("Basic distribution validation complete.")
            
            # ------- Additional Dynamic Validation -------
            # Validate trial counts match configuration
            actual_counts = all_designs.groupby('trial_type').size().to_dict()
            validation_results['trial_counts'] = {
                'expected': self.trial_counts,
                'actual_per_participant': {k: v/len(designs) for k, v in actual_counts.items()}
            }
            
            # Validate SOA distribution
            soa_distribution = all_designs.groupby(['trial_type', 'soa_value_ms']).size().to_dict()
            validation_results['soa_distribution'] = soa_distribution
            
            # Find the approximate experiment duration
            if all_designs.shape[0] > 0:
                # Get last timestamp from each participant
                participant_durations = []
                for participant_id, group in all_designs.groupby('participant_id'):
                    if not group.empty:
                        last_time = group['timestamp_after_jitter'].iloc[-1]
                        parts = last_time.split(':')
                        if len(parts) == 2:
                            minutes = int(parts[0])
                            seconds = float(parts[1])
                            total_seconds = minutes * 60 + seconds
                            participant_durations.append(total_seconds)
                
                if participant_durations:
                    avg_duration = sum(participant_durations) / len(participant_durations)
                    validation_results['estimated_duration'] = {
                        'seconds': avg_duration,
                        'minutes': avg_duration / 60,
                        'per_trial': avg_duration / (sum(self.trial_counts.values()) / len(designs))
                    }
                    print(f"Estimated experiment duration: {avg_duration/60:.1f} minutes")
            
            # Save detailed validation report to JSON
            validation_report_path = os.path.join(self.output_dir, 'dynamic_validation_report.json')
            try:
                with open(validation_report_path, 'w') as f:
                    json.dump(validation_results, f, indent=4, default=str)
                print(f"Validation report saved to: {validation_report_path}")
            except Exception as e:
                print(f"Warning: Error saving validation report: {str(e)}")
            
            # Print summary of validation results
            print("\nValidation Results:")
            print(f"Total participants: {len(designs)}")
            print(f"Total trials per participant: {sum(self.trial_counts.values())}")
            print("Trial counts by condition:")
            for trial_type, count in self.trial_counts.items():
                actual = actual_counts.get(trial_type, 0) / len(designs)
                print(f"  {trial_type}: {count} expected, {actual:.1f} actual")
            
            if 'estimated_duration' in validation_results:
                print(f"Estimated duration: {validation_results['estimated_duration']['minutes']:.1f} minutes")
            
        except Exception as e:
            print(f"Error during validation: {str(e)}")
            import traceback
            traceback.print_exc()
            print("Validation could not be completed. Returning partial results.")
        
        print("Validation complete.")
        return validation_results

# ======================================================================
# AUDIO GENERATOR
# ======================================================================

class DynamicAudioGenerator:
    """
    Generates participant-specific audio files with embedded stimuli,
    dynamically adjusting audio length based on required trials.
    """
    
    def __init__(self, base_audio_path, output_dir, looming_stimuli_paths, tactile_stimulus_path, timestamps_csv_path=None):
        """
        Initialize the audio generator.
        
        Args:
            base_audio_path: Path to the base box breathing audio
            output_dir: Directory to save participant-specific audio files
            looming_stimuli_paths: Dictionary of paths to looming stimuli sounds
            tactile_stimulus_path: Path to the tactile stimulus sound
            timestamps_csv_path: Path to CSV with breathing timestamps
        """
        # Store stimulus paths
        self.looming_stimuli = looming_stimuli_paths
        self.tactile_stimulus_path = tactile_stimulus_path
        
        # Load base audio (breathing exercise)
        self.base_audio_data, self.sample_rate = sf.read(base_audio_path)
        
        # Store base audio info for dynamic extension
        self.base_audio_duration = len(self.base_audio_data) / self.sample_rate
        
        if EXPERIMENT_CONFIG['debug_mode']:
            print(f"Base audio duration: {self.base_audio_duration:.1f} seconds "
                  f"({self.base_audio_duration/60:.1f} minutes)")
        
        # Load tactile stimulus
        self.tactile_data, _ = sf.read(tactile_stimulus_path)
        
        # Load timestamps for breathing cycles
        self.timestamps_csv_path = timestamps_csv_path
        if timestamps_csv_path:
            self.breathing_analyzer = BreathingCycleAnalyzer(timestamps_csv_path)
        else:
            self.breathing_analyzer = None
        
        # Output directory for audio files
        self.output_dir = output_dir
        os.makedirs(output_dir, exist_ok=True)
    
    def extract_loopable_audio_segment(self):
        """
        Extract the loopable segment from the base audio.
        This extracts the audio between holds in the breathing cycle.
        
        Returns:
            Loopable audio segment as numpy array
        """
        if self.breathing_analyzer is None:
            # If no analyzer, return the whole audio
            return self.base_audio_data
        
        # Get the first few timestamps to identify a loopable segment
        if len(self.breathing_analyzer.start_timestamps) >= 2:
            # Get the first and second cycle start points
            first_cycle_start_ms = self.breathing_analyzer.start_timestamps[0]
            second_cycle_start_ms = self.breathing_analyzer.start_timestamps[1]
            
            # Convert to samples
            first_start_sample = int(first_cycle_start_ms / 1000 * self.sample_rate)
            second_start_sample = int(second_cycle_start_ms / 1000 * self.sample_rate)
            
            # Extract the loopable segment
            loopable_segment = self.base_audio_data[first_start_sample:second_start_sample]
            
            if EXPERIMENT_CONFIG['debug_mode']:
                loop_duration = len(loopable_segment) / self.sample_rate
                print(f"Extracted loopable segment: {loop_duration:.2f} seconds")
            
            return loopable_segment
        else:
            # Fall back to using the whole audio
            return self.base_audio_data
    
    def create_custom_length_audio(self, required_cycles):
        """
        Create a custom length audio file by repeating loopable segments.
        
        Args:
            required_cycles: Number of breathing cycles required
            
        Returns:
            Custom length audio array
        """
        # Extract loopable segment
        loopable_segment = self.extract_loopable_audio_segment()
        
        # For very short requirements, just return the base audio
        if required_cycles <= 1:
            return self.base_audio_data.copy()
        
        # Calculate how many repetitions we need
        if self.breathing_analyzer and self.breathing_analyzer.cycle_duration_ms > 0:
            # Calculate based on segment length and cycle duration
            segment_duration = len(loopable_segment) / self.sample_rate
            cycle_duration = self.breathing_analyzer.cycle_duration_ms / 1000
            
            # Ensure we have at least one complete cycle
            if segment_duration < cycle_duration:
                repeats_needed = required_cycles
            else:
                cycles_per_segment = segment_duration / cycle_duration
                repeats_needed = math.ceil(required_cycles / cycles_per_segment)
        else:
            # If no analyzer or invalid cycle duration, use a simple approach
            repeats_needed = required_cycles
        
        if EXPERIMENT_CONFIG['debug_mode']:
            print(f"Creating custom audio: needs {required_cycles} cycles, "
                  f"using {repeats_needed} repetitions of loopable segment")
        
        # Combine loopable segment with beginning and ending
        if repeats_needed <= 1:
            # No extension needed
            return self.base_audio_data.copy()
        else:
            # Create looped middle section
            middle_section = np.tile(loopable_segment, (repeats_needed, 1))
            
            # Get introduction and conclusion sections
            intro_samples = len(loopable_segment)  # Use same length as loopable segment
            outro_samples = min(len(self.base_audio_data) - intro_samples, intro_samples)
            
            intro_section = self.base_audio_data[:intro_samples]
            outro_section = self.base_audio_data[-outro_samples:]
            
            # Combine sections
            try:
                combined_audio = np.vstack([intro_section, middle_section, outro_section])
                return combined_audio
            except ValueError:
                # If shapes don't match (e.g., different channel counts), just use the looped middle
                if EXPERIMENT_CONFIG['debug_mode']:
                    print("Warning: Error combining audio sections. Using only the looped middle.")
                return middle_section
    
    def generate_participant_audio(self, participant_id, design_df):
        """
        Generate two synchronized audio files for a participant.
        
        Args:
            participant_id: Participant identifier
            design_df: DataFrame with stimuli design
                
        Returns:
            Tuple (looming_audio, tactile_audio, injection_log)
        """
        # Check if design is empty
        if design_df.empty:
            print(f"Warning: Empty design for participant {participant_id}. Skipping audio generation.")
            return None, None, pd.DataFrame()
        
        # Calculate required cycles based on trial count
        total_trials = len(design_df)
        required_cycles = 0
        
        if self.breathing_analyzer:
            required_cycles = self.breathing_analyzer.get_cycles_for_trials(total_trials)
        else:
            # Estimate based on timestamps
            if not design_df.empty:
                # Approximate using timestamp format
                last_timestamp = design_df['timestamp_with_soa'].iloc[-1]
                time_parts = last_timestamp.split(':')
                if len(time_parts) == 2:
                    minutes = int(time_parts[0])
                    seconds = float(time_parts[1])
                    last_time_seconds = minutes * 60 + seconds
                    # Estimate cycles (assuming ~16 sec per cycle)
                    required_cycles = math.ceil(last_time_seconds / 16) + 1
        
        if EXPERIMENT_CONFIG['debug_mode']:
            print(f"Required cycles for {total_trials} trials: {required_cycles}")
        
        # Create custom length audio base
        custom_audio = self.create_custom_length_audio(required_cycles)
        
        # Check if audio generation succeeded
        if custom_audio is None:
            print(f"Error creating custom audio for participant {participant_id}")
            return None, None, pd.DataFrame()
        
        # Create two separate audio streams
        looming_audio = custom_audio.copy()  # Box breathing with looming sounds
        
        # Create tactile audio with same shape as looming audio
        if len(custom_audio.shape) > 1:  # Stereo
            tactile_audio = np.zeros_like(custom_audio)  # Empty stereo track
        else:  # Mono
            tactile_audio = np.zeros(len(custom_audio))  # Empty mono track
        
        # Ensure tactile stimulus is compatible format
        if self.tactile_data.ndim == 1 and looming_audio.ndim > 1:  # If tactile is mono and looming is stereo
            tactile_stereo = np.column_stack((self.tactile_data, self.tactile_data))
        elif self.tactile_data.ndim > 1 and looming_audio.ndim == 1:  # If tactile is stereo and looming is mono
            tactile_stereo = np.mean(self.tactile_data, axis=1)
        else:
            tactile_stereo = self.tactile_data
            
        # Initialize log to track all stimulus injections
        injection_log = []
        
        # Verify required columns
        required_columns = ['stimulus_type', 'jittered_ms', 'soa_ms', 'soa_value_ms']
        missing_columns = [col for col in required_columns if col not in design_df.columns]
        if missing_columns:
            print(f"Warning: Design DataFrame missing required columns: {missing_columns}. "
                  f"Cannot ensure precise synchronization.")
            return None, None, pd.DataFrame()
        
        # Process each trial in the design
        for _, row in design_df.iterrows():
            # Get exact millisecond timestamps from the design CSV
            jittered_ms = row['jittered_ms']  # For looming stimuli
            soa_ms = row['soa_ms']  # For tactile stimuli
            
            # Convert to sample indices
            looming_injection_point = int(jittered_ms / 1000 * self.sample_rate)
            tactile_injection_point = int(soa_ms / 1000 * self.sample_rate)
            
            # Get SOA value for logging
            soa_value_ms = row['soa_value_ms']
            
            # Load the appropriate looming stimulus based on design
            stim_type = row['stimulus_type']
            looming_path = self.looming_stimuli.get(stim_type)
            if looming_path is None:
                continue  # Skip if stimulus type not found
            
            # Read the looming stimulus audio file
            looming_data, _ = sf.read(looming_path)
            
            # Check if stimuli fit within the audio bounds
            looming_fits = (looming_injection_point + len(looming_data) <= len(looming_audio))
            
            if looming_fits:
                # Inject looming stimulus into the first audio track
                try:
                    if looming_audio.ndim > 1 and looming_data.ndim > 1:  # Both stereo
                        looming_audio[looming_injection_point:looming_injection_point+len(looming_data)] += looming_data
                    elif looming_audio.ndim > 1 and looming_data.ndim == 1:  # Looming stereo, data mono
                        for ch in range(looming_audio.shape[1]):
                            looming_audio[looming_injection_point:looming_injection_point+len(looming_data), ch] += looming_data
                    elif looming_audio.ndim == 1 and looming_data.ndim > 1:  # Looming mono, data stereo
                        looming_audio[looming_injection_point:looming_injection_point+len(looming_data)] += np.mean(looming_data, axis=1)
                    else:  # Both mono
                        looming_audio[looming_injection_point:looming_injection_point+len(looming_data)] += looming_data
                except ValueError as e:
                    print(f"Error injecting looming stimulus: {e}, shapes: "
                          f"looming_audio={looming_audio.shape}, looming_data={looming_data.shape}")
                    continue
                
                # Inject tactile stimulus for non-catch trials
                if row.get('is_tactile', True) and row['trial_type'] != 'catch':
                    tactile_fits = (tactile_injection_point >= 0 and 
                                  tactile_injection_point + len(tactile_stereo) <= len(tactile_audio))
                    
                    if tactile_fits:
                        try:
                            if tactile_audio.ndim > 1 and tactile_stereo.ndim > 1:  # Both stereo
                                tactile_audio[tactile_injection_point:tactile_injection_point+len(tactile_stereo)] += tactile_stereo
                            elif tactile_audio.ndim > 1 and tactile_stereo.ndim == 1:  # Tactile stereo, data mono
                                for ch in range(tactile_audio.shape[1]):
                                    tactile_audio[tactile_injection_point:tactile_injection_point+len(tactile_stereo), ch] += tactile_stereo
                            elif tactile_audio.ndim == 1 and tactile_stereo.ndim > 1:  # Tactile mono, data stereo
                                tactile_audio[tactile_injection_point:tactile_injection_point+len(tactile_stereo)] += np.mean(tactile_stereo, axis=1)
                            else:  # Both mono
                                tactile_audio[tactile_injection_point:tactile_injection_point+len(tactile_stereo)] += tactile_stereo
                        except ValueError as e:
                            print(f"Error injecting tactile stimulus: {e}, shapes: "
                                  f"tactile_audio={tactile_audio.shape}, tactile_stereo={tactile_stereo.shape}")
                            continue
                    
                # Log details of this injection
                injection_log.append({
                    'participant_id': participant_id,
                    'trial_number': row.get('trial_number', 0),
                    'stimulus_type': stim_type,
                    'soa_condition_idx': row.get('soa_condition_idx', 0),
                    'soa_value_ms': soa_value_ms,
                    'looming_injection_sample': looming_injection_point,
                    'tactile_injection_sample': tactile_injection_point,
                    'looming_time_ms': looming_injection_point / self.sample_rate * 1000,
                    'tactile_time_ms': tactile_injection_point / self.sample_rate * 1000,
                    'jittered_ms': jittered_ms,
                    'soa_ms': soa_ms,
                    'is_tactile': row.get('is_tactile', True)
                })
        
        return looming_audio, tactile_audio, pd.DataFrame(injection_log)
    
    def save_participant_files(self, participant_id, looming_audio, tactile_audio, injection_log, design_filename):
        """
        Save participant-specific audio and log files.
        
        Args:
            participant_id: Participant identifier
            looming_audio: Modified audio with looming stimuli
            tactile_audio: Audio track with tactile stimuli
            injection_log: DataFrame with injection details
            design_filename: Base design filename for naming consistency
        
        Returns:
            Tuple of saved file paths
        """
        # Check if audio data is None
        if looming_audio is None or tactile_audio is None:
            print(f"Warning: Cannot save audio for participant {participant_id} - missing audio data")
            return None, None, None
        
        # Use the design CSV base name (remove .csv extension)
        base_name = os.path.splitext(design_filename)[0]
        
        # Define filenames for output files
        looming_filename = f"{base_name}_looming.wav"
        tactile_filename = f"{base_name}_tactile.wav"
        log_filename = f"{base_name}_stimuli_log.csv"
        
        # Create full paths
        looming_path = os.path.join(self.output_dir, looming_filename)
        tactile_path = os.path.join(self.output_dir, tactile_filename)
        log_path = os.path.join(self.output_dir, log_filename)
        
        # Save the audio files
        sf.write(looming_path, looming_audio, self.sample_rate)
        sf.write(tactile_path, tactile_audio, self.sample_rate)
        
        # Save the detailed injection log (for analysis)
        if not injection_log.empty:
            injection_log.to_csv(log_path, index=False)
        
        audio_duration = len(looming_audio) / self.sample_rate
        print(f"Looming audio saved: {looming_filename} (duration: {audio_duration:.1f} seconds)")
        print(f"Tactile audio saved: {tactile_filename}")
        print(f"Injection log saved: {log_filename}")
        
        return looming_path, tactile_path, log_path


# ======================================================================
# MAIN FUNCTION
# ======================================================================

def main():
    """
    Main function to run the dynamic PPS experiment setup process.
    """
    # Ensure output directories exist
    os.makedirs(DESIGN_OUTPUT_DIR, exist_ok=True)
    os.makedirs(AUDIO_OUTPUT_DIR, exist_ok=True)
    
    # Calculate and display trial counts
    trial_counts = calculate_trial_counts()
    total_trials = sum(trial_counts.values())
    
    # Display configuration overview
    print("\n===== PPS EXPERIMENT GENERATOR =====")
    print(f"Repetitions per condition: {EXPERIMENT_CONFIG['repetitions_per_condition']}")
    
    # Show which conditions are included
    active_conditions = [cond for cond, include in EXPERIMENT_CONFIG['included_conditions'].items() if include]
    print("Included conditions:", ", ".join(active_conditions))
    
    print(f"Catch trial percentage: {EXPERIMENT_CONFIG['catch_trial_percentage']}%")
    print(f"Total trials per participant: {total_trials}")
    print(f"Output directory: {DESIGN_OUTPUT_DIR}")
    
    # Step 1: Generate counterbalanced designs
    print("\n===== STEP 1: GENERATING COUNTERBALANCED DESIGNS =====")
    print(f"Using SOA conditions (ms): {SOA_CONDITIONS_MS}")
    
    counterbalancer = DynamicCounterbalancer(
        output_dir=DESIGN_OUTPUT_DIR,
        soa_conditions=SOA_CONDITIONS_MS,
        jitter_options=JITTER_OPTIONS_MS
    )
    
    designs = counterbalancer.generate_comprehensive_design(
        num_participants=NUM_PARTICIPANTS, 
        reference_csv_path=TIMESTAMPS_CSV_PATH
    )
    
    # Save designs to CSV files
    counterbalancer.save_designs(designs)
    
    # Run validation checks
    print("\n===== RUNNING VALIDATION CHECKS =====")
    validation_results = counterbalancer.validate_designs(designs)
    
    # Step 2: Generate participant-specific audio files
    print("\n===== STEP 2: GENERATING TRULY DYNAMIC AUDIO FILES =====")
    print(f"Creating TWO audio files per participant:")
    print(f"  - Looming audio: Box breathing with directional looming sounds")
    print(f"  - Tactile audio: Synchronized tactile stimuli with SOA offsets")
    
    # Initialize the audio generator
    audio_generator = DynamicAudioGenerator(
        base_audio_path=BASE_AUDIO_PATH,
        output_dir=AUDIO_OUTPUT_DIR,
        looming_stimuli_paths=LOOMING_STIMULI,
        tactile_stimulus_path=TACTILE_STIMULUS_PATH,
        timestamps_csv_path=TIMESTAMPS_CSV_PATH
    )
    
    # Process each design and generate audio pairs
    for i, design_df in enumerate(designs):
        if design_df.empty:
            continue
            
        participant_id = design_df['participant_id'].iloc[0]
        design_filename = f"participant_{participant_id}_design.csv"
        
        print(f"\nProcessing participant {participant_id} ({i+1}/{NUM_PARTICIPANTS})...")
        
        # Verify design CSV has necessary columns
        if 'jittered_ms' not in design_df.columns or 'soa_ms' not in design_df.columns:
            print(f"Warning: Design for participant {participant_id} missing critical timing columns.")
            print("Reading design from saved CSV file to ensure correct timing...")
            
            # Re-read the design CSV
            design_file_path = os.path.join(DESIGN_OUTPUT_DIR, design_filename)
            if os.path.exists(design_file_path):
                design_df = pd.read_csv(design_file_path)
                print("Successfully loaded design from CSV with all timing information.")
            else:
                print(f"Error: Cannot find design file {design_file_path}")
                continue
        
        # Generate audio for this participant
        try:
            looming_audio, tactile_audio, injection_log = audio_generator.generate_participant_audio(
                participant_id, 
                design_df
            )
            
            # Save both audio files and the log
            audio_generator.save_participant_files(
                participant_id, 
                looming_audio, 
                tactile_audio, 
                injection_log, 
                design_filename
            )
            
            # Verify synchronization
            print("Verifying synchronization...")
            if not injection_log.empty:
                for i, row in injection_log.head(3).iterrows():
                    looming_time = row['looming_time_ms']
                    tactile_time = row['tactile_time_ms']
                    expected_diff = row['soa_value_ms']
                    actual_diff = tactile_time - looming_time
                    
                    print(f"  Trial {i+1}: SOA expected={expected_diff}ms, actual={actual_diff:.2f}ms, " +
                          f"difference={abs(actual_diff-expected_diff):.4f}ms")
                
                if len(injection_log) > 3:
                    print(f"  ... and {len(injection_log) - 3} more trials")
            else:
                print("  Warning: No stimuli were injected - check design parameters")
                
        except Exception as e:
            print(f"Error generating audio for participant {participant_id}: {str(e)}")
            import traceback
            traceback.print_exc()
            continue
    
    # Print summary
    print(f"\n===== GENERATION COMPLETE =====")
    print(f"All designs generated and saved to: {DESIGN_OUTPUT_DIR}")
    print(f"All audio files generated and saved to: {AUDIO_OUTPUT_DIR}")
    print(f"Created {NUM_PARTICIPANTS} participant designs with:")
    for trial_type, count in trial_counts.items():
        print(f"  - {count} {trial_type} trials")
    
    if 'estimated_duration' in validation_results:
        print(f"Estimated experiment duration: {validation_results['estimated_duration']['minutes']:.1f} minutes per participant")
    
    print("Each participant has TWO audio files: one with looming stimuli and one with tactile stimuli")
    print("Process complete.")

if __name__ == "__main__":
    main()
```
